### 3.7.4. Distributionally Robust Optimization

$$
\min_{\mathbf{x}}\; \sup_{\mathbb{P}\in\mathcal{P}}\; \mathbb{E}_{\mathbb{P}}\big[\, Q(\mathbf{x},\boldsymbol\xi)\,\big],
$$
with ambiguity set $\mathcal{P}$ of distributions sharing known moments.

**Explanation:**

Distributionally robust optimization (DRO) hedges against not knowing the true distribution: it optimizes the worst-case expectation over an *ambiguity set* $\mathcal{P}$ of plausible distributions, typically all distributions with a given mean and variance. A distributional chance constraint $\inf_{\mathbb{P}\in\mathcal{P}}\Pr[D\le s]\ge 1-\varepsilon$ then admits a closed form through the one-sided Chebyshev (Cantelli) bound, giving a margin $\sigma\sqrt{(1-\varepsilon)/\varepsilon}$ instead of the Gaussian quantile used in [chance-constrained optimization](./02_chance_constrained_optimization.ipynb). Because it trusts only the moments, DRO is more conservative than the Gaussian model.

**Numerical Example:**

Guarantee $\Pr[D\le s]\ge 0.95$ using *only* the mean $\mu=100$ and standard deviation $\sigma=15$, for **any** distribution with those moments. The Cantelli bound $\Pr[D-\mu\ge t]\le \tfrac{\sigma^2}{\sigma^2+t^2}$ forces
$$
\frac{\sigma^2}{\sigma^2 + t^2} \le 0.05
\;\Rightarrow\;
t \ge \sigma\sqrt{\tfrac{1-\varepsilon}{\varepsilon}} = 15\sqrt{19} \approx 65.4,
$$
so $s^\star \approx 165.4$. This far exceeds the Gaussian requirement $s=124.7$: distributional robustness costs an extra $40.7$-unit buffer for immunity to the distribution's shape.

```python
import sympy as sp

mean, standard_deviation, epsilon = 100, 15, sp.Rational(5, 100)
cantelli_margin = standard_deviation * sp.sqrt((1 - epsilon) / epsilon)
distributionally_robust_level = mean + cantelli_margin
gaussian_level = 124.7

print("Cantelli margin t =", round(float(cantelli_margin), 1))
print("DRO stock level s =", round(float(distributionally_robust_level), 1))
print("extra buffer over Gaussian model =", round(float(distributionally_robust_level) - gaussian_level, 1))
```

**Equivalent numpy implementation:**

```python
import numpy as np

mean, standard_deviation, epsilon = 100.0, 15.0, 0.05
cantelli_margin = standard_deviation * np.sqrt((1 - epsilon) / epsilon)
distributionally_robust_level = mean + cantelli_margin

print("Cantelli margin t =", round(cantelli_margin, 1))
print("DRO stock level s =", round(distributionally_robust_level, 1))
```
