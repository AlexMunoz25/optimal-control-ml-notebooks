### 3.2.7.7. Optimal Detector Design


$$P_{ij}=\operatorname{prob}(\hat y=i\mid y=j)$$


**Explanation:**

Detector design chooses a rule that maps observations to decisions. Randomized detectors are represented by probability matrices, so the feasible detector set is convex. Optimality usually minimizes a linear risk or satisfies linear performance constraints.

**Properties:**
- Detector probabilities are nonnegative.
- Each true-hypothesis column sums to one.
- Average error is a linear function of the detector matrix.

**Intuition:**

The example computes the average error of a two-hypothesis randomized detector.


**Numerical Example:**

Let
$$
P=\begin{bmatrix}9/10&1/5\\1/10&4/5\end{bmatrix}
$$
be the detection probability matrix. The error under hypothesis $1$ is
$$
P_{21}=1/10,
$$
and the error under hypothesis $2$ is
$$
P_{12}=1/5.
$$

With equal priors, the average error is
$$
\frac12\cdot\frac1{10}+\frac12\cdot\frac15
=\frac1{20}+\frac1{10}=\frac3{20}.
$$


```python
import sympy as sp

detection_matrix = sp.Matrix([[sp.Rational(9, 10), sp.Rational(1, 5)], [sp.Rational(1, 10), sp.Rational(4, 5)]])
average_error = sp.Rational(1, 2) * detection_matrix[1, 0] + sp.Rational(1, 2) * detection_matrix[0, 1]

average_error
```
