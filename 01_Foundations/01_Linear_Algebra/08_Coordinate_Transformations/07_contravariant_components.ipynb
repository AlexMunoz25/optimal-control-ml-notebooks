### 1.1.8.7. Contravariant Components


$$
\mathbf{A} = A^i \mathbf{e}_i,
\qquad
A^i = \mathbf{A} \cdot \mathbf{e}^{\,i}
$$


**Explanation:**

In Fleisch's treatment, contravariant components are obtained by parallel projection onto the coordinate axes. They are the coefficients that rebuild the vector directly from the original basis vectors, so this notebook links to the [dual basis](./06_dual_basis.ipynb) instead of rederiving the coefficient rule. The coordinate-change law is opposite to the basis change, which explains the term contravariant.

<p align="center">
  <img src="../../../Figures/01010803_contravariant_components_basis.jpeg" alt="Contravariant components in a skew basis" width="420">
</p>

**Example:**

For a vector in a skew basis, the contravariant components are the coefficients in the expansion $\mathbf{A} = A^1 \mathbf{e}_1 + A^2 \mathbf{e}_2$.


```python
import sympy as sp

basis_matrix = sp.Matrix([[1, sp.Rational(1, 2)], [0, sp.sqrt(3) / 2]])
vector_a = sp.Matrix([3, 2])
contravariant_components = sp.simplify(basis_matrix.LUsolve(vector_a))

print("contravariant_components =", contravariant_components)
```
