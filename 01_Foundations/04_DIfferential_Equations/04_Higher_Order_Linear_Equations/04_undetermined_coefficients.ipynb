### 1.4.4.4. Undetermined Coefficients

$$
y = y_c + y_p, \qquad
y_p = \text{(trial form matching } g(x)\text{) with coefficients fixed by substitution} .
$$

**Explanation:**

When the forcing $g(x)$ is a polynomial, exponential, sine/cosine, or a product/sum of these, a particular solution $y_p$ can be guessed as a trial form of the *same type* with undetermined coefficients, which are then pinned down by substituting into the equation and matching like terms. The general solution is $y = y_c + y_p$, with $y_c$ from the [characteristic equation](./03_homogeneous_equations_with_constant_coefficients.ipynb). If a trial term already appears in $y_c$, it must be multiplied by $x$ (or $x^2$) before substituting. The method is fast but limited to these "nice" forcing functions; general forcing needs [variation of parameters](./05_variation_of_parameters.ipynb).

**Properties:**
- Trial forms: polynomial $\to$ polynomial; $e^{\alpha x} \to A e^{\alpha x}$; $\sin\beta x$ or $\cos\beta x \to A\cos\beta x + B\sin\beta x$.
- By the superposition principle, a sum of forcing terms is handled term by term.

**Numerical Example:**

Solve $y'' + 4y' - 2y = 2x^2 - 3x + 6$.

**Step 1 — complementary function.** Auxiliary $m^2 + 4m - 2 = 0$ gives $m = -2 \pm \sqrt{6}$, so

$$
y_c = c_1 e^{(-2-\sqrt6)x} + c_2 e^{(-2+\sqrt6)x}.
$$

**Step 2 — trial form.** Since $g$ is a quadratic, try $y_p = Ax^2 + Bx + C$, with $y_p' = 2Ax + B$, $y_p'' = 2A$. Substituting,

$$
y_p'' + 4y_p' - 2y_p = -2Ax^2 + (8A - 2B)x + (2A + 4B - 2C) = 2x^2 - 3x + 6 .
$$

**Step 3 — match coefficients.** Equating like powers of $x$:

$$
-2A = 2, \qquad 8A - 2B = -3, \qquad 2A + 4B - 2C = 6 ,
$$

which solve to $A = -1$, $B = -\tfrac52$, $C = -9$. Hence $y_p = -x^2 - \tfrac52 x - 9$ and

$$
y = y_c + y_p = c_1 e^{(-2-\sqrt6)x} + c_2 e^{(-2+\sqrt6)x} - x^2 - \tfrac52 x - 9 .
$$

```python
import sympy as sp

x = sp.symbols("x")
y = sp.Function("y")
A, B, C = sp.symbols("A B C")

trial = A * x**2 + B * x + C
residual = trial.diff(x, 2) + 4 * trial.diff(x) - 2 * trial - (2 * x**2 - 3 * x + 6)
coefficients = sp.solve(sp.Poly(residual, x).all_coeffs(), [A, B, C])
particular = trial.subs(coefficients)

general = sp.dsolve(sp.Eq(y(x).diff(x, 2) + 4 * y(x).diff(x) - 2 * y(x), 2 * x**2 - 3 * x + 6), y(x))

print("matched coefficients:", coefficients)
print("particular y_p =", particular)
print("general solution:", general.rhs)
```
