### 3.2.2.1. Affine Sets and Convex Sets


$$\theta x_1+(1-\theta)x_2\in C,\qquad0\leq\theta\leq1$$


**Explanation:**

Affine and convex sets are defined by closure under combinations of points. Affine combinations allow signed coefficients that sum to one; convex combinations also require nonnegative coefficients. This distinction separates lines and planes from segments, polygons, and other feasible regions built from averages.

The expression $\theta x_1+(1-\theta)x_2$ is the weighted average of two [vectors](../../../01_Foundations/01_Linear_Algebra/02_Vector/01_vector_definition.ipynb). The condition $0\leq\theta\leq1$ keeps the point on the line segment between $x_1$ and $x_2$, and convexity requires that point to remain in $C$.

**Properties:**
- Affine combinations require coefficients to sum to one.
- Convex combinations also require nonnegative coefficients.
- Every affine set is convex, but a line segment is convex without being affine.

**Numerical Example:**

The figure above shows the same segment geometry; the arithmetic computes one point on that segment.

Let
$$
x_1=\begin{bmatrix}0\\0\end{bmatrix},\quad
x_2=\begin{bmatrix}4\\2\end{bmatrix},\quad
\theta=\frac14.
$$

The convex combination is
$$
\theta x_1+(1-\theta)x_2=\frac14\begin{bmatrix}0\\0\end{bmatrix}+\frac34\begin{bmatrix}4\\2\end{bmatrix}.
$$

Coordinate by coordinate,
$$
\frac14\cdot0+\frac34\cdot4=3,
\qquad
\frac14\cdot0+\frac34\cdot2=\frac32.
$$

Therefore the point is
$$
\begin{bmatrix}3\\3/2\end{bmatrix}.
$$


```python
import sympy as sp

point_1 = sp.Matrix([0, 0])
point_2 = sp.Matrix([4, 2])
theta = sp.Rational(1, 4)
combination = theta * point_1 + (1 - theta) * point_2

combination
```
