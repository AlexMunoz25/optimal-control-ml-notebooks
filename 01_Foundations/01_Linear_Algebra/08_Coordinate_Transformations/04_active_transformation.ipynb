### 1.1.8.4. Active Transformation


$$
\begin{aligned}
\mathbf{A}' &= R_{\mathrm{active}}(\alpha)\,\mathbf{A}, \\
R_{\mathrm{active}}(\alpha)
&=
\begin{bmatrix}
\cos \alpha & -\sin \alpha \\
\sin \alpha & \cos \alpha
\end{bmatrix}, \\
A_x' &= A_x \cos \alpha - A_y \sin \alpha,
\qquad
A_y' = A_x \sin \alpha + A_y \cos \alpha
\end{aligned}
$$


**Explanation:**

An active transformation rotates the vector itself while the coordinate axes stay fixed. The same direct matrix used to rotate basis vectors also rotates any vector expressed in that basis. This differs from a [passive transformation](./03_passive_transformation.ipynb), where the vector is unchanged and only its coordinates are updated.

  <img src="../../../Figures/01010801_vector_rotation_active.jpeg" alt="Active rotation of a vector in a fixed basis" width="580">
</p>

**Example:**

For $\theta=150^\circ$,

$$
R=\begin{bmatrix}-\sqrt3/2&-1/2\\1/2&-\sqrt3/2\end{bmatrix}.
$$

Apply it to

$$
\mathbf{a}=\begin{bmatrix}5\\3\end{bmatrix}:
$$

$$
R\mathbf{a}
=\begin{bmatrix}(-\sqrt3/2)5+(-1/2)3\\(1/2)5+(-\sqrt3/2)3\end{bmatrix}
=\begin{bmatrix}-5\sqrt3/2-3/2\\5/2-3\sqrt3/2\end{bmatrix}.
$$


```python
import sympy as sp

rotation_angle = sp.rad(150)
active_transformation_matrix = sp.Matrix([
    [sp.cos(rotation_angle), -sp.sin(rotation_angle)],
    [sp.sin(rotation_angle), sp.cos(rotation_angle)],
])
original_vector_components = sp.Matrix([5, 3])
rotated_vector_components = active_transformation_matrix * original_vector_components

print("active_transformation_matrix =")
sp.pprint(active_transformation_matrix)
print("rotated_vector_components =")
sp.pprint(rotated_vector_components)
```
