### 1.1.5.11. Coordinates with Respect to a Basis


$$
\mathbf{v} = c_1 \mathbf{b}_1 + c_2 \mathbf{b}_2 + \cdots + c_n \mathbf{b}_n
$$

$$
\left[ \mathbf{v} \right]_B =
\begin{bmatrix}
c_1 \\
c_2 \\
\vdots \\
c_n
\end{bmatrix}
$$


**Explanation:**

Coordinates are the coefficients of a vector relative to a chosen [basis](./10_basis.ipynb). Orthonormal, orthogonal, and generic bases all describe the same vector space, but they compute those coefficients in different ways. The same coefficient viewpoint later reappears in [contravariant components](../08_Coordinate_Transformations/07_contravariant_components.ipynb).

**Properties:**
- In an orthonormal basis, each coordinate is a dot product with a basis vector.
- In a generic basis, the coordinates are obtained by solving a linear system.

**Example:**

If

$$
\mathbf{v} =
\begin{bmatrix}
5 \\
3
\end{bmatrix},
\qquad
B =
\left\{
\begin{bmatrix}
1 \\
1
\end{bmatrix},
\begin{bmatrix}
1 \\
-1
\end{bmatrix}
\right\},
$$

then solving for the coefficients gives

$$
\left[ \mathbf{v} \right]_B =
\begin{bmatrix}
4 \\
1
\end{bmatrix}.
$$


```python
import sympy as sp

basis_matrix = sp.Matrix([[1, 1], [1, -1]])
vector_v = sp.Matrix([5, 3])
coordinate_vector = basis_matrix.LUsolve(vector_v)

print("coordinate_vector =", coordinate_vector)
print("reconstructed_vector =", basis_matrix * coordinate_vector)
```
