### 1.1.7.20. Complex Vector Space


$$
\mathbb{C}^n
$$


**Explanation:**

A complex vector space is obtained by allowing complex scalars in the vector-space axioms. This changes the underlying field while leaving the linear structure intact, and it sets up the conjugation rules used in [Hermitian transpose](./22_hermitian_transpose.ipynb) and [complex inner product](./23_complex_inner_product.ipynb).

**Example:**

A vector such as

$$
\begin{bmatrix}
1 + i \\
2 - i
\end{bmatrix}
$$

belongs to $\mathbb{C}^2$.


```python
import sympy as sp

vector_z = sp.Matrix([1 + sp.I, 2 - sp.I])
scalar_multiple = sp.expand((2 - sp.I) * vector_z)

print("vector_z =")
sp.pprint(vector_z.T)
print("scalar_multiple =")
sp.pprint(scalar_multiple.T)
```
