### 3.2.6.2. Least-Squares, L1, and Chebyshev Approximation


$$\|r\|_2^2,\qquad\|r\|_1,\qquad\|r\|_\infty$$


**Explanation:**

Least-squares, l1, and Chebyshev approximation use different convex residual measures. Squared two-norm emphasizes larger errors, one-norm sums absolute errors, and infinity norm controls the worst residual.

The vector $r$ is the residual. The three expressions measure it with squared Euclidean norm, one-norm, and infinity norm, so the same residual can penalize total squared error, total absolute error, or worst absolute error.

**Properties:**
- The same residual vector can have different objective values.
- The infinity norm is the maximum absolute residual.
- Norm and penalty choices encode robustness or smoothness preferences.

**Intuition:**

The example evaluates three common residual measures on one vector.


**Numerical Example:**

Let
$$
r=\begin{bmatrix}1\\-2\\3\end{bmatrix}.
$$

The squared two-norm is
$$
\|r\|_2^2=1^2+(-2)^2+3^2=1+4+9=14.
$$

The one-norm is
$$
\|r\|_1=|1|+|-2|+|3|=1+2+3=6.
$$

The infinity norm is
$$
\|r\|_\infty=\max\{1,2,3\}=3.
$$


```python
import sympy as sp

residual = sp.Matrix([1, -2, 3])
squared_l2 = sum(entry ** 2 for entry in residual)
l1_norm = sum(abs(entry) for entry in residual)
linfinity_norm = max(abs(entry) for entry in residual)

squared_l2, l1_norm, linfinity_norm
```
