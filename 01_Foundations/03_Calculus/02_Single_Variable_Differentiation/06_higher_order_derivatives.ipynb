### 1.3.2.6. Higher-Order Derivatives

$$
f''(x) = \frac{d}{dx}\big(f'(x)\big), \qquad
f^{(n)}(x) = \frac{d^{n}}{dx^{n}} f(x) = \frac{d}{dx}\big(f^{(n-1)}(x)\big).
$$

**Explanation:**

Higher-order derivatives are obtained by differentiating repeatedly: the second derivative $f''$ is the rate of change of the slope and measures curvature, the third is the rate of change of curvature, and so on. The second derivative drives the [concavity / second-derivative test](./09_convexity_and_second_derivative_test.ipynb) for classifying extrema and supplies the curvature term in [Taylor approximation](../04_Sequences_Series_and_Taylor_Approximation/04_taylor_and_maclaurin_series.ipynb); its multivariable analogue is the [Hessian](../05_Multivariable_Differential_Calculus/09_hessian_matrix.ipynb). In dynamics, position differentiated once gives velocity and twice gives acceleration.

**Properties:**
- For a degree-$n$ polynomial, $f^{(n)}$ is constant and $f^{(n+1)} = 0$.
- $\frac{d^n}{dx^n} e^{x} = e^{x}$ for all $n$, and the derivatives of $\sin$/$\cos$ cycle with period $4$.

**Numerical Example:**

For $f(x) = x^4 - 3x^2 + 5$, differentiate successively:

$$
f'(x) = 4x^3 - 6x, \quad f''(x) = 12x^2 - 6, \quad f'''(x) = 24x, \quad f^{(4)}(x) = 24, \quad f^{(5)}(x) = 0 .
$$

The fourth derivative is the constant $24 = 4!$, and all further derivatives vanish. Evaluating curvature at $x = 1$: $f''(1) = 12 - 6 = 6 > 0$, so the graph is concave up there.

```python
import sympy as sp

x = sp.symbols("x")
f = x**4 - 3 * x**2 + 5

successive_derivatives = [sp.diff(f, x, order) for order in range(1, 6)]
curvature_at_one = successive_derivatives[1].subs(x, 1)

for order, derivative in enumerate(successive_derivatives, start=1):
    print(f"f^({order})(x) =", derivative)
print("f''(1) =", curvature_at_one, "-> concave up" if curvature_at_one > 0 else "")
```

**Equivalent CasADi implementation:**

Repeatedly applying `ca.gradient` to its own output produces the successive derivatives as new graphs.

```python
import casadi as ca

x = ca.SX.sym("x")
f = x**4 - 3 * x**2 + 5

current = f
for order in range(1, 5):
    current = ca.gradient(current, x)
    print(f"f^({order})(x) =", ca.simplify(current))
```
