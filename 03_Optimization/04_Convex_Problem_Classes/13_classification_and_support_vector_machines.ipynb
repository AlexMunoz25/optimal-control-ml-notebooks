### 3.4.13. Classification and Support Vector Machines

$$a^Tx_i-b\geq1,
\qquad a^Ty_i-b\leq-1$$


**Explanation:**

Linear classification separates labeled point sets with an affine hyperplane. Margin constraints are linear inequalities in the classifier parameters, so the feasible classifier set is convex. Feasibility means every sample satisfies its signed margin condition.

The vector $a$ is the classifier normal and $b$ is the offset. The inequalities require samples $x_i$ and $y_i$ to lie on opposite sides of the separating hyperplane with margin at least one.

**Properties:**
- Positive samples must lie on one side of the hyperplane.
- Negative samples must lie on the other side.
- Margins measure signed [distance](../../01_Foundations/01_Linear_Algebra/05_Geometrical_Aspects_of_Linear_Algebra/04_distance_between_points.ipynb) up to scaling.

**Intuition:**

The figure above shows affine discrimination; the calculation checks one positive and one negative margin.

<img src="../../Figures/030319_bv_fig_8_8_linear_discrimination.jpeg" alt="Boyd and Vandenberghe Figure 8.8: affine classifier separating two point sets" style="width: 60%; height: auto;">


**Numerical Example:**

The figure above shows a separating affine classifier. Let
$$
a=\begin{bmatrix}1\\0\end{bmatrix},\qquad b=0.
$$

For a positive point $x_+=(2,0)$,
$$
a^Tx_+-b=2\geq1.
$$

For a negative point $x_-=(-2,0)$,
$$
a^Tx_--b=-2\leq-1.
$$

Both margin constraints are satisfied.


```python
import sympy as sp

normal_vector = sp.Matrix([1, 0])
positive_point = sp.Matrix([2, 0])
negative_point = sp.Matrix([-2, 0])
positive_margin = (normal_vector.T * positive_point)[0]
negative_margin = (normal_vector.T * negative_point)[0]

print("positive_margin =", positive_margin)
print("negative_margin =", negative_margin)
print("positive_margin >= 1 =", positive_margin >= 1)
print("negative_margin <= -1 =", negative_margin <= -1)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

normal_vector = ca.DM([1, 0])
positive_point = ca.DM([2, 0])
negative_point = ca.DM([-2, 0])
positive_margin = ca.dot(normal_vector, positive_point)
negative_margin = ca.dot(normal_vector, negative_point)

print("positive_margin =", positive_margin)
print("negative_margin =", negative_margin)
print("positive_margin >= 1 =", bool(positive_margin >= 1))
print("negative_margin <= -1 =", bool(negative_margin <= -1))
```

**Support vector machines — soft margins:**

Hard-margin separation is infeasible whenever the two point sets overlap. The support vector machine relaxes each margin with a nonnegative slack variable $u_i$,

$$
y_i\,(a^{T}x_i-b)\;\ge\;1-u_i,\qquad u_i\ge 0,
$$

and trades separation against margin width by minimizing $\tfrac12\|a\|_2^2+C\sum_i u_i$. The slack $u_i$ is the amount by which sample $i$ violates the unit margin, so the smallest feasible slack for a point whose score is $a^{T}x-b=\tfrac12$ is $u=\tfrac12$. The penalized objective is a [quadratic program](06_quadratic_programming_and_qcqp.ipynb), which is why support vector classification is a workhorse convex model in machine learning.
