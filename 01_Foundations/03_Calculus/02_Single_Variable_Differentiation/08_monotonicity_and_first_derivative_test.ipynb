### 1.3.2.8. Monotonicity and the First Derivative Test

$$
f'(x) > 0 \text{ on } I \Rightarrow f \text{ increasing on } I, \qquad
f'(x) < 0 \text{ on } I \Rightarrow f \text{ decreasing on } I .
$$

$$
\text{At a critical point } c\ (f'(c)=0):\quad
f' : + \to -\ \Rightarrow\ \text{local max}, \qquad
f' : - \to +\ \Rightarrow\ \text{local min}.
$$

**Explanation:**

The sign of the first derivative tells whether a function is increasing or decreasing, so the points where $f'$ changes sign are the local extrema. A *critical point* is where $f'(c) = 0$ or $f'$ is undefined; by Fermat's theorem every local extremum of a differentiable function occurs at a critical point, making critical points the candidate set. The first derivative test classifies each candidate by how the sign of $f'$ changes across it. This is the one-dimensional core of the [necessary optimality conditions](../../../03_Optimization/01_Optimization_Fundamentals/04_necessary_optimality_conditions.ipynb) that optimization generalizes to the gradient.

**Properties:**
- $f'(c) = 0$ is necessary but not sufficient for a local extremum (e.g. $f(x) = x^3$ at $0$).
- A sign change $+ \to -$ marks a local maximum; $- \to +$ marks a local minimum; no change means no extremum.

**Intuition:**

Where the derivative changes sign the graph turns around: a local maximum where $f'$ goes from positive to negative and a local minimum where it goes from negative to positive.

**Numerical Example:**

Let $f(x) = x^3 - 3x$. Then $f'(x) = 3x^2 - 3 = 3(x-1)(x+1)$, so the critical points are $x = -1$ and $x = 1$. Testing the sign of $f'$ on the three intervals:

$$
(-\infty, -1):\ f' > 0\ (\text{incr}), \quad (-1, 1):\ f' < 0\ (\text{decr}), \quad (1, \infty):\ f' > 0\ (\text{incr}).
$$

At $x = -1$ the sign goes $+ \to -$: a **local maximum**, $f(-1) = 2$. At $x = 1$ the sign goes $- \to +$: a **local minimum**, $f(1) = -2$.

```python
import sympy as sp

x = sp.symbols("x")
f = x**3 - 3 * x
derivative = sp.diff(f, x)
critical_points = sp.solve(derivative, x)

test_points = [-2, 0, 2]
sign_table = [(point, sp.sign(derivative.subs(x, point))) for point in test_points]
classification = {
    point: ("local max" if sp.diff(f, x, 2).subs(x, point) < 0 else "local min")
    for point in critical_points
}

print("f'(x) =", sp.factor(derivative))
print("critical points:", critical_points)
for point, sign in sign_table:
    print(f"  sign f'({point}) = {sign}")
for point, label in classification.items():
    print(f"x = {point}: {label}, f = {f.subs(x, point)}")
```

**Equivalent CasADi implementation:**

CasADi forms $f'$ as a graph; its roots (the critical points) are found with a root-finder, and the sign of $f'$ on either side classifies them.

```python
import casadi as ca

x = ca.SX.sym("x")
f = x**3 - 3 * x
derivative_function = ca.Function("df", [x], [ca.gradient(f, x)])
root_finder = ca.rootfinder("r", "newton", {"x": x, "g": ca.gradient(f, x)})

left_root = float(root_finder(x0=-2)["x"])
right_root = float(root_finder(x0=2)["x"])
print("critical points:", round(left_root, 6), round(right_root, 6))
print("f' just left/right of", round(left_root, 3), ":",
      float(derivative_function(left_root - 0.1)), float(derivative_function(left_root + 0.1)))
```
