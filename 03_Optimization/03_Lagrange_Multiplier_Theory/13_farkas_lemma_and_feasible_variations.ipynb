### 3.3.13. Farkas Lemma and Feasible Variations

$$
c^T y\le 0 \text{ for all } y \text{ with } a_j^T y\le 0
\quad \Longleftrightarrow \quad
c=\sum_{j=1}^{r}\mu_j a_j,\ \mu_j\ge 0
$$

**Explanation:**

Farkas' lemma connects two views of the same cone geometry. A vector $c$ has nonpositive [inner product](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/02_inner_product_space.ipynb) with every direction satisfying the linear inequalities exactly when $c$ lies in the cone generated by the corresponding normals. In multiplier theory, this is the bridge from “no descent direction in the first-order feasible cone” to the existence of nonnegative multipliers.

The figures show the generated cone and how it supplies multipliers for linear constraints at a local minimum.

**Numerical Example:**

Let

$$
a_1=\begin{bmatrix}1\\0\end{bmatrix},
\qquad
a_2=\begin{bmatrix}0\\1\end{bmatrix}.
$$

The cone generated by $a_1$ and $a_2$ is the nonnegative quadrant. Choose

$$
c=\begin{bmatrix}2\\3\end{bmatrix}.
$$

Then

$$
c=2a_1+3a_2,
$$

with multipliers

$$
\mu_1=2\ge 0,
\qquad
\mu_2=3\ge 0.
$$

Now take any $y$ satisfying

$$
a_1^T y=y_1\le 0,
\qquad
a_2^T y=y_2\le 0.
$$

Then

$$
c^T y=2y_1+3y_2\le 0
$$

because both terms are nonpositive. This verifies the forward direction of Farkas' lemma in a concrete cone.

```python
import sympy as sp

mu_1, mu_2, y_1, y_2 = sp.symbols('mu_1 mu_2 y_1 y_2')

generator_1 = sp.Matrix([1, 0])
generator_2 = sp.Matrix([0, 1])
vector_c = sp.Matrix([2, 3])
multiplier_solution = sp.solve(vector_c - mu_1 * generator_1 - mu_2 * generator_2, [mu_1, mu_2], dict=True)[0]
y = sp.Matrix([y_1, y_2])
inner_product = vector_c.dot(y)

result = {
    'multiplier_solution': multiplier_solution,
    'inner_product': inner_product,
    'inner_product_at_example_direction': inner_product.subs({y_1: -4, y_2: -5}),
}

for label, value in result.items():
    print(f"{label} =", value)
```
