### 1.3.3.1. Antiderivative and Indefinite Integral

$$
F'(x) = f(x) \;\Longleftrightarrow\; \int f(x)\,dx = F(x) + C .
$$

**Explanation:**

An antiderivative of $f$ is any function $F$ whose [derivative](../02_Single_Variable_Differentiation/01_derivative_definition.ipynb) is $f$. Antiderivatives are not unique: if $F$ works, so does $F + C$ for any constant $C$, because the derivative of a constant is zero, and the Mean Value Theorem shows these are the *only* antiderivatives. The whole family is written as the indefinite integral $\int f(x)\,dx = F(x) + C$. Reversing each differentiation rule gives a corresponding integration formula, and recovering a quantity from its rate of change — position from velocity, cost from marginal cost — is exactly this operation.

**Properties:**
- Power rule: $\int x^{n}\,dx = \dfrac{x^{n+1}}{n+1} + C$ for $n \neq -1$, and $\int x^{-1}\,dx = \ln|x| + C$.
- Linearity: $\int \big(\alpha f + \beta g\big)\,dx = \alpha\!\int f\,dx + \beta\!\int g\,dx$.

**Numerical Example:**

Find the antiderivative of $f(x) = 3x^2 + 2x$. Reversing the power rule term by term,

$$
\int (3x^2 + 2x)\,dx = 3\cdot\frac{x^3}{3} + 2\cdot\frac{x^2}{2} + C = x^3 + x^2 + C .
$$

Check by differentiating: $\dfrac{d}{dx}(x^3 + x^2 + C) = 3x^2 + 2x = f(x)$. To single out one antiderivative, impose a condition such as $F(0) = 5$, giving $C = 5$ and $F(x) = x^3 + x^2 + 5$.

```python
import sympy as sp

x = sp.symbols("x")
f = 3 * x**2 + 2 * x

general_antiderivative = sp.integrate(f, x)
recovered = sp.diff(general_antiderivative, x)

constant = sp.symbols("C")
particular = general_antiderivative + constant
constant_value = sp.solve(sp.Eq(particular.subs(x, 0), 5), constant)[0]

print("indefinite integral :", general_antiderivative, "+ C")
print("d/dx of result       :", recovered)
print("with F(0)=5 -> C =", constant_value, ": F(x) =", general_antiderivative + constant_value)
```
