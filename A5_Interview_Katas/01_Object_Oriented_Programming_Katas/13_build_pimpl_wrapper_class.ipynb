### A5.1.13. Build PIMPL Wrapper Class

**Problem:**

Build a class that hides its implementation details behind a private inner class. The public interface delegates all work to the inner class, so internal changes do not affect external code.

**Explanation:**

PIMPL (Pointer to Implementation) separates interface from implementation. The public class holds a reference to a private implementation object. All methods on the public class forward to the implementation. This means the implementation can change freely without affecting any code that depends on the public class.

How to build it:

1. Define an inner class (conventionally named `_Impl`) that holds all the state and logic.
2. The outer class creates an instance of `_Impl` in `__init__`.
3. Every public method on the outer class delegates to the corresponding method on `self._impl`.
4. External code only sees the outer class. The inner class can be refactored without changing the outer interface.

**Example:**

A `Widget` class delegates to `_WidgetImpl`. The implementation stores coordinates and draw logic. The public `Widget` simply forwards calls.

```python
class _WidgetImpl:
    def __init__(self, x_pos, y_pos, label):
        self.x_pos = x_pos
        self.y_pos = y_pos
        self.label = label

    def draw(self):
        return f"Drawing '{self.label}' at ({self.x_pos}, {self.y_pos})"

    def move(self, dx, dy):
        self.x_pos += dx
        self.y_pos += dy


class Widget:
    def __init__(self, x_pos, y_pos, label):
        self._impl = _WidgetImpl(x_pos, y_pos, label)

    def draw(self):
        return self._impl.draw()

    def move(self, dx, dy):
        self._impl.move(dx, dy)


widget = Widget(10, 20, "Button")
print(widget.draw())

widget.move(5, -3)
print(widget.draw())

print(f"Public interface: {[m for m in dir(widget) if not m.startswith('_')]}")
```
