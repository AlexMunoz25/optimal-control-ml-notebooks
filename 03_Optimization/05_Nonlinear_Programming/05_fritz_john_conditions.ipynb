### 3.5.5. Fritz John Conditions

$$
\mu_0^*\nabla f(x^*)+
\sum_{i=1}^{m}\lambda_i^*\nabla h_i(x^*)+
\sum_{j=1}^{r}\mu_j^*\nabla g_j(x^*)=0
$$

$$
\mu_0^*\ge 0, \qquad \mu_j^*\ge 0, \qquad
(\mu_0^*,\lambda^*,\mu^*)\neq 0
$$

**Explanation:**

The Fritz John conditions are necessary conditions that do not require regularity. They introduce a nonnegative multiplier $\mu_0^*$ on the objective gradient; when $\mu_0^*>0$, the conditions can be normalized into ordinary KKT conditions. When $\mu_0^*=0$, the constraints alone create an abnormal stationarity relation, which is exactly what can happen at nonregular points.

**Intuition:**

The nonregular two-constraint example from the equality section becomes an inequality example by replacing one circle equality with an inside constraint and the other with an outside constraint. The geometry still forces the only feasible point to be the origin, but standard KKT multipliers fail while Fritz John multipliers remain available.

**Numerical Example:**

Use the inequality version of the nonregular example:

$$
f(x)=x_1+x_2,
$$

$$
g_1(x)=(x_1-1)^2+x_2^2-1\le 0,
\qquad
g_2(x)=-(x_1-2)^2-x_2^2+4\le 0.
$$

At the only feasible point $x^*=(0,0)$,

$$
\nabla f(x^*)=\begin{bmatrix}1\\1\end{bmatrix},
\quad
\nabla g_1(x^*)=\begin{bmatrix}-2\\0\end{bmatrix},
\quad
\nabla g_2(x^*)=\begin{bmatrix}4\\0\end{bmatrix}.
$$

The Fritz John stationarity condition is

$$
\mu_0\begin{bmatrix}1\\1\end{bmatrix}
+\mu_1\begin{bmatrix}-2\\0\end{bmatrix}
+\mu_2\begin{bmatrix}4\\0\end{bmatrix}
=0.
$$

The second coordinate gives

$$
\mu_0=0.
$$

Then the first coordinate gives

$$
-2\mu_1+4\mu_2=0
\quad \Rightarrow \quad
\mu_1=2\mu_2.
$$

Choose $\mu_2=1$. Then

$$
\mu_0=0, \qquad \mu_1=2, \qquad \mu_2=1,
$$

which is nonzero and has nonnegative inequality multipliers. This is an abnormal Fritz John multiplier set, not a normalized KKT multiplier set.

```python
import sympy as sp

mu_0, mu_1, mu_2 = sp.symbols('mu_0 mu_1 mu_2')

gradient_cost = sp.Matrix([1, 1])
gradient_constraint_1 = sp.Matrix([-2, 0])
gradient_constraint_2 = sp.Matrix([4, 0])
stationarity = mu_0 * gradient_cost + mu_1 * gradient_constraint_1 + mu_2 * gradient_constraint_2
solution_family = sp.solve(stationarity, [mu_0, mu_1], dict=True)[0]
example_multipliers = {mu_0: 0, mu_1: 2, mu_2: 1}

result = {
    'solution_family': solution_family,
    'example_multipliers': example_multipliers,
    'stationarity_at_example': stationarity.subs(example_multipliers),
}

for label, value in result.items():
    print(f"{label} =", value)
```
