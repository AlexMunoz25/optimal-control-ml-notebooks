### 1.1.5.2. Plane in R^3


$$
P = \left\{ (x, y, z) \in \mathbb{R}^3 \mid Ax + By + Cz = D \right\}
$$

$$
\mathbf{n} \cdot \left( \mathbf{x} - \mathbf{p}_0 \right) = 0
$$

$$
\mathbf{x} = \mathbf{p}_0 + s \mathbf{v} + t \mathbf{w}
$$


**Explanation:**

A plane in $\mathbb{R}^3$ is the two-dimensional set of points orthogonal to a normal vector $\mathbf{n}$, or equivalently the set generated by two independent in-plane directions. Savov presents the algebraic, geometric, and parametric equations as interchangeable views of the same plane, so the notebook keeps those formulas together instead of splitting the concept artificially. This notebook depends on the idea of a [line](./01_line_in_r3.ipynb) and supports later work on [distances](./06_distance_from_plane_to_origin.ipynb) and [plane projections](./08_projection_onto_a_plane.ipynb).

**Properties:**
- Scaling the coefficients $A$, $B$, $C$, and $D$ by a nonzero constant leaves the plane unchanged.
- The normal vector determines the orientation of the plane, not its size.

**Example:**

If

$$
\mathbf{p}_0 =
\begin{bmatrix}
1 \\
0 \\
0
\end{bmatrix},
\qquad
\mathbf{n} =
\begin{bmatrix}
1 \\
1 \\
1
\end{bmatrix},
$$

then

$$
\mathbf{n} \cdot (\mathbf{x} - \mathbf{p}_0) = 0
\quad \Longrightarrow \quad
x + y + z = 1.
$$


```python
import sympy as sp

plane_normal = sp.Matrix([1, 1, 1])
point_in_plane = sp.Matrix([1, 0, 0])
symbol_x, symbol_y, symbol_z = sp.symbols("symbol_x symbol_y symbol_z")
plane_equation = sp.expand(plane_normal.dot(sp.Matrix([symbol_x, symbol_y, symbol_z]) - point_in_plane))

print("plane_normal =", list(plane_normal))
print("point_in_plane =", list(point_in_plane))
print("plane_equation =", sp.Eq(plane_equation, 0))
```
