### 3.1.23. Quasi-Newton Methods


$$
p^k=x^{k+1}-x^k,
\qquad
q^k=\nabla f(x^{k+1})-\nabla f(x^k),
\qquad
D^{k+1}q^k=p^k.
$$


**Explanation:**

Quasi-Newton methods try to obtain much of Newton's local speed without computing the Hessian explicitly. After a step has been taken, the change in position $p^k$ and the change in gradient $q^k$ contain curvature information. The secant relation $D^{k+1}q^k=p^k$ asks the inverse-Hessian approximation to be consistent with that observed curvature. If $D^k$ remains [positive definite](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/17_positive_definite_matrix.ipynb), then $-D^k\nabla f(x^k)$ is a descent direction.

The vector $p^k$ is the step between two iterates, and $q^k$ is the corresponding gradient change. The secant equation $D^{k+1}q^k=p^k$ says the new inverse-Hessian approximation should reproduce the observed curvature along that step.

The central idea is learning curvature from motion. Steepest descent uses no curvature, Newton computes curvature directly, and quasi-Newton methods update an approximation from successive gradients. These methods naturally sit beside conjugate gradient because both seek better directions than steepest descent, but they make different storage and computation tradeoffs.

**Intuition:**

For a quadratic function, the gradient changes exactly according to the Hessian times the step, so the secant relation is exact. For a nonquadratic function, it is a local approximation that becomes more informative near a solution. The code cell checks the exact secant behavior on a small quadratic model.


**Numerical Example:**

For a quadratic cost with
$$
Q=\begin{bmatrix}3&1\\1&2\end{bmatrix},
\qquad \nabla f(x)=Qx,
$$
choose
$$
x^k=\begin{bmatrix}0\\0\end{bmatrix},
\qquad x^{k+1}=\begin{bmatrix}1\\-1\end{bmatrix}.
$$

The position increment is
$$
p^k=x^{k+1}-x^k=\begin{bmatrix}1\\-1\end{bmatrix}.
$$

The gradients are
$$
\nabla f(x^k)=\begin{bmatrix}0\\0\end{bmatrix},
\qquad
\nabla f(x^{k+1})=Qx^{k+1}
=\begin{bmatrix}2\\-1\end{bmatrix}.
$$

Thus
$$
q^k=\nabla f(x^{k+1})-\nabla f(x^k)=\begin{bmatrix}2\\-1\end{bmatrix}.
$$

Since
$$
Qp^k=\begin{bmatrix}3&1\\1&2\end{bmatrix}\begin{bmatrix}1\\-1\end{bmatrix}
=\begin{bmatrix}2\\-1\end{bmatrix}=q^k,
$$
the secant relation is exact for this quadratic.


```python
import casadi as ca

decision_variable = ca.SX.sym("x", 2)
Q = ca.DM([[3, 1], [1, 2]])
objective = 0.5 * ca.bilin(Q, decision_variable, decision_variable)
hessian, gradient_expr = ca.hessian(objective, decision_variable)
gradient_function = ca.Function("grad", [decision_variable], [gradient_expr])
hessian_matrix = ca.Function("hessian_eval", [decision_variable], [hessian])(ca.DM([0, 0]))

previous_point = ca.DM([0, 0])
next_point = ca.DM([1, -1])
previous_gradient = gradient_function(previous_point)
next_gradient = gradient_function(next_point)
position_increment = next_point - previous_point
gradient_increment = next_gradient - previous_gradient
secant_residual = gradient_increment - hessian_matrix @ position_increment

print("previous_gradient =", previous_gradient)
print("next_gradient =", next_gradient)
print("position_increment =", position_increment)
print("gradient_increment =", gradient_increment)
print("secant_residual =", secant_residual)
```
