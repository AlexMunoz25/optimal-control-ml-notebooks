### 1.1.6.7. Invertible Linear Transformation


$$
T^{-1}(T(\mathbf{v})) = \mathbf{v}
$$

$$
\operatorname{Ker}(T) = \{ \mathbf{0} \}
\Longleftrightarrow
T \text{ is invertible}
$$


**Explanation:**

An invertible linear transformation can be undone by a reverse transformation. For linear maps, invertibility is tied to a trivial [kernel](./03_kernel.ipynb), because no nonzero direction can collapse to zero. The full matrix-side equivalences are collected in the [invertible matrix theorem](./14_invertible_matrix_theorem.ipynb).

**Example:**

A stretch matrix

$$
\begin{bmatrix}
2 & 0 \\
0 & 1
\end{bmatrix}
$$

is invertible because no nonzero vector is sent to zero, and the inverse stretch is

$$
\begin{bmatrix}
\frac{1}{2} & 0 \\
0 & 1
\end{bmatrix}.
$$


```python
import sympy as sp

invertible_matrix = sp.Matrix([[2, 0], [0, 1]])
inverse_matrix = invertible_matrix.inv()

print("inverse_matrix =")
sp.pprint(inverse_matrix)
print("product_matrix =")
sp.pprint(invertible_matrix * inverse_matrix)
```
