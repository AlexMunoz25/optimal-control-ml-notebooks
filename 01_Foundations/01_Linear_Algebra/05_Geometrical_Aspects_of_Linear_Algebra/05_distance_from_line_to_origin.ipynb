### 1.1.5.5. Distance from a Line to the Origin


$$
d(\ell, \mathbf{0}) =
\left\lVert
\mathbf{p}_0 -
\frac{\mathbf{p}_0 \cdot \mathbf{v}}{\lVert \mathbf{v} \rVert^2} \mathbf{v}
\right\rVert
$$


**Explanation:**

Savov derives the distance from a line to the origin by removing the component of the anchor point $\mathbf{p}_0$ that lies along the line direction $\mathbf{v}$. The remaining perpendicular component has exactly the length of the shortest segment from the origin to the line. This notebook depends on a [line in $\mathbb{R}^3$](./01_line_in_r3.ipynb) and on the [projection onto a line](./07_projection_onto_a_line.ipynb), so those concepts are referenced rather than repeated.

**Example:**

If

$$
\ell = \left\{
\begin{bmatrix}
4 \\
5 \\
6
\end{bmatrix}
+ t
\begin{bmatrix}
1 \\
0 \\
1
\end{bmatrix}
\mid t \in \mathbb{R}
\right\},
$$

then

$$
d(\ell, \mathbf{0}) =
\left\lVert
\begin{bmatrix}
4 \\
5 \\
6
\end{bmatrix}
- 5
\begin{bmatrix}
1 \\
0 \\
1
\end{bmatrix}
\right\rVert
=
\left\lVert
\begin{bmatrix}
-1 \\
5 \\
1
\end{bmatrix}
\right\rVert
= 3 \sqrt{3}.
$$


```python
import sympy as sp

line_point = sp.Matrix([4, 5, 6])
line_direction = sp.Matrix([1, 0, 1])
parallel_component = line_point.dot(line_direction) / line_direction.dot(line_direction) * line_direction
perpendicular_component = line_point - parallel_component
distance_value = sp.simplify(perpendicular_component.norm())

print("parallel_component =", parallel_component)
print("perpendicular_component =", perpendicular_component)
print("distance_value =", distance_value)
```
