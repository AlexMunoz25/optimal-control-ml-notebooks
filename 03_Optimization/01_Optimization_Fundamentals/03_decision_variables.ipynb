### 3.1.3. Decision Variables

$$
\mathbf{x} = (x_1,\dots,x_n)\in\mathbb{R}^n,
\qquad
\min_{\mathbf{x}} \; f(\mathbf{x};\,\mathbf{p}),
$$
with fixed problem data (parameters) $\mathbf{p}$.

**Explanation:**

The decision variables are the quantities the optimizer is free to choose; collecting them into the vector $\mathbf{x}\in\mathbb{R}^n$ fixes the dimension of the search space. They are distinct from *parameters*, which are fixed data of the problem: a parameter is held constant while $\mathbf{x}$ varies. The way a quantity enters the model тАФ as a variable or as data тАФ determines the geometry of the [feasible set](./04_feasible_sets.ipynb) and the shape of the [objective](./02_objective_functions.ipynb).

**Numerical Example:**

Take the parametrized objective
$$
f(\mathbf{x};\,\mathbf{p}) = (x_1 - p_1)^2 + (x_2 - p_2)^2,
\qquad \mathbf{p} = (3, 4).
$$

Only $\mathbf{x}$ is chosen; $\mathbf{p}$ is fixed data. Setting the gradient with respect to the decision variables to zero,
$$
\nabla_{\mathbf{x}} f = \begin{bmatrix} 2(x_1 - 3) \\ 2(x_2 - 4)\end{bmatrix} = \mathbf 0
\;\Rightarrow\;
\mathbf{x}^\star = (3, 4),
$$
so the minimizer tracks the parameter. Differentiating instead with respect to $\mathbf{p}$ would be meaningless here тАФ the parameter is not a degree of freedom.

```python
import sympy as sp

x_1, x_2 = sp.symbols("x_1 x_2")
p_1, p_2 = sp.symbols("p_1 p_2")
decision_variable = sp.Matrix([x_1, x_2])
parameters = {p_1: 3, p_2: 4}
objective = (x_1 - p_1)**2 + (x_2 - p_2)**2

gradient = sp.Matrix([sp.diff(objective, variable) for variable in decision_variable])
stationary = sp.solve(gradient, [x_1, x_2], dict=True)[0]
minimizer = decision_variable.subs(stationary).subs(parameters)

print("gradient w.r.t. decision variables ="); sp.pprint(gradient)
print("minimizer for p = (3, 4) ="); sp.pprint(minimizer)
```

**Equivalent casadi implementation:**

```python
import casadi as ca

decision_variable = ca.SX.sym("x", 2)
parameter = ca.SX.sym("p", 2)
objective_expression = (decision_variable[0] - parameter[0])**2 + (decision_variable[1] - parameter[1])**2

solver = ca.nlpsol("solver", "ipopt", {"x": decision_variable, "p": parameter, "f": objective_expression},
                   {"ipopt.print_level": 0, "print_time": 0})
solution = solver(x0=[0, 0], p=[3, 4])

print("minimizer for p = (3, 4) =", solution["x"])
```
