### 3.8.3. Bayesian Optimization

$$
\mathbf{x}_{k+1} = \arg\max_{\mathbf{x}}\; \alpha\big(\mathbf{x}\,;\,\mathcal{D}_k\big),
\qquad
\alpha = \mu(\mathbf{x}) \ \text{traded off against}\ \sigma(\mathbf{x}),
$$
with surrogate mean $\mu$ and uncertainty $\sigma$ fit to data $\mathcal{D}_k$.

**Explanation:**

Bayesian optimization minimizes an expensive black-box objective by fitting a cheap probabilistic *surrogate* to the observations and choosing the next evaluation where an *acquisition function* is largest. The acquisition balances exploitation (sampling where the surrogate predicts low values) against exploration (sampling where the surrogate is uncertain), so no gradient of the true objective is ever needed. It is the method of choice when each evaluation is costly, as in hyperparameter tuning or physical experiments.

**Intuition:**

The surrogate interpolates the sampled points with a confidence band; the next sample is placed where the acquisition — high uncertainty plus promising mean — peaks.

![Surrogate model and acquisition function](../../Figures/030803_bayesian_optimization_surrogate.png)

**Numerical Example:**

Minimize $f(x) = \sin(x) + 0.1x$ on $[0,10]$ having observed it at $x\in\{1,4,7,9\}$. A distance-weighted surrogate predicts a mean $\mu(x)$ and an uncertainty $\sigma(x)$ that vanishes at observed points and grows between them. The next query maximizes the acquisition
$$
\alpha(x) = \sigma(x) - \big(\mu(x) - \min \mu\big),
$$
which favors points that are both uncertain and predicted to be low. Evaluated on a grid, the acquisition peaks near $x\approx 4.7$, so that becomes the next point to sample.

```python
import numpy as np

grid = np.linspace(0, 10, 400)
sample_x = np.array([1.0, 4.0, 7.0, 9.0])
sample_y = np.sin(sample_x) + 0.1 * sample_x

length_scale = 1.2
weights = np.exp(-0.5 * ((grid[:, None] - sample_x[None, :]) / length_scale) ** 2)
surrogate_mean = weights @ sample_y / weights.sum(axis=1)
surrogate_uncertainty = 0.6 * (1 - weights.max(axis=1))
acquisition = surrogate_uncertainty - (surrogate_mean - surrogate_mean.min())

next_query = grid[int(np.argmax(acquisition))]
print("next query point x =", round(float(next_query), 2))
print("surrogate mean there =", round(float(surrogate_mean[int(np.argmax(acquisition))]), 3))
```
