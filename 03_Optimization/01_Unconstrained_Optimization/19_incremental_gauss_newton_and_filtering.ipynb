### 3.1.19. Incremental Gauss-Newton and Filtering


$$
\begin{aligned}
K_k &= P_{k-1}H_k^{\mathsf T}\left(H_kP_{k-1}H_k^{\mathsf T}+R_k\right)^{-1}, \\
\hat{x}_k &= \hat{x}_{k-1}+K_k\left(y_k-H_k\hat{x}_{k-1}\right), \\
P_k &= \left(I-K_kH_k\right)P_{k-1}.
\end{aligned}
$$


**Explanation:**

Incremental Gauss-Newton methods extend the block-by-block idea to second-order least-squares approximations. In the linear case, the same algebra leads to recursive least squares and Kalman-filter-like formulas, where the estimate and an information or covariance matrix are updated as each measurement arrives. This material shows that optimization algorithms and filtering recursions are closely related when both are viewed through least squares.

The display uses filtering notation. $K_k$ is the gain matrix, $P_{k-1}$ and $P_k$ are covariance-like matrices, $H_k$ maps the current state estimate to the measurement space, and the update adds a gain-weighted residual $y_k-H_k\hat x_{k-1}$ to the estimate.

The main pedagogical point is that batch and recursive formulations can represent the same estimation problem. A batch method collects all residuals and solves one large normal equation; an incremental method folds in information one block at a time. For nonlinear residuals, linearization gives the extended Kalman filter viewpoint, with the usual caution that the quality of the update depends on the quality of the local linear model.

**Intuition:**

A single recursive least squares update uses one measurement row and one residual. The gain matrix determines how much the estimate should move in response to the new discrepancy. The formula uses [matrix transpose](../../01_Foundations/01_Linear_Algebra/03_Matrix/06_matrix_transpose.ipynb) and [matrix inverse](../../01_Foundations/01_Linear_Algebra/03_Matrix/11_matrix_inverse.ipynb), but avoids rebuilding the entire batch system after each new datum.


**Numerical Example:**

Use one recursive least-squares measurement with
$$
\hat{x}_0=\begin{bmatrix}0\\0\end{bmatrix},
\quad
P_0=10I,
\quad
H=\begin{bmatrix}1&2\end{bmatrix},
\quad
y=3,
\quad
R=0.5.
$$

The scalar denominator is
$$
HP_0H^{\mathsf T}+R
=\begin{bmatrix}1&2\end{bmatrix}
\begin{bmatrix}10&0\\0&10\end{bmatrix}
\begin{bmatrix}1\\2\end{bmatrix}+0.5
=10+40+0.5=50.5.
$$

The gain is
$$
K=P_0H^{\mathsf T}(HP_0H^{\mathsf T}+R)^{-1}
=\frac{1}{50.5}\begin{bmatrix}10\\20\end{bmatrix}
\approx\begin{bmatrix}0.1980\\0.3960\end{bmatrix}.
$$

The residual is
$$
y-H\hat{x}_0=3-0=3,
$$
so the updated estimate is
$$
\hat{x}_1=\hat{x}_0+K(3)
\approx\begin{bmatrix}0.5941\\1.1881\end{bmatrix}.
$$


```python
import numpy as np

estimate = np.array([0.0, 0.0])
information_matrix = 10.0 * np.eye(2)
measurement_matrix = np.array([[1.0, 2.0]])
measurement = np.array([3.0])
measurement_variance = np.array([[0.5]])
kalman_gain = information_matrix @ measurement_matrix.T @ np.linalg.inv(
    measurement_matrix @ information_matrix @ measurement_matrix.T + measurement_variance
)
updated_estimate = estimate + kalman_gain @ (measurement - measurement_matrix @ estimate)
updated_information_matrix = (np.eye(2) - kalman_gain @ measurement_matrix) @ information_matrix

print("updated_estimate =", updated_estimate)
print("updated_information_matrix =", updated_information_matrix)
```
