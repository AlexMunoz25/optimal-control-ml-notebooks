### 3.2.5.4. Duality Geometry and Saddle Points


$$L(x^*,\lambda,\nu)\leq L(x^*,\lambda^*,\nu^*)\leq L(x,\lambda^*,\nu^*)$$


**Explanation:**

A saddle point of the Lagrangian links primal and dual optimality in one inequality. At such a point, the Lagrangian is minimized over primal variables and maximized over dual variables, so the primal and dual values meet.

The middle term evaluates the Lagrangian at a primal-dual candidate pair. The left inequality says changing multipliers cannot increase it at $x^*$, and the right inequality says changing the primal variable cannot decrease it at the optimal multipliers.

**Properties:**
- Inequality multipliers must be nonnegative.
- The dual function is an infimum over primal variables.
- The lower bound can be loose unless the multiplier is optimal.

**Intuition:**

The calculation minimizes one scalar Lagrangian exactly.

**Numerical Example:**

The figure above gives the visual version of the same calculation.

Consider the Lagrangian
$$
L(x,2)=x^2+2(x-1)=x^2+2x-2.
$$

Differentiate with respect to $x$:
$$
\frac{d}{dx}L(x,2)=2x+2.
$$

Set this to zero:
$$
2x+2=0 \quad\Longrightarrow\quad x=-1.
$$

The dual-function value at this multiplier is
$$
L(-1,2)=(-1)^2+2(-1-1)=1-4=-3.
$$

This is a valid lower bound for the associated minimization problem.


```python
import sympy as sp

variable = sp.symbols('x')
lagrangian = variable ** 2 + 2 * (variable - 1)
stationary_point = sp.solve(sp.diff(lagrangian, variable), variable)[0]
dual_value = sp.simplify(lagrangian.subs(variable, stationary_point))

print("stationary_point =", stationary_point)
print("dual_value =", dual_value)
```
