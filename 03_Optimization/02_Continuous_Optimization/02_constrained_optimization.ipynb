### 3.2.2. Constrained Optimization

$$
\begin{aligned}
\min_{\mathbf{x}\in\mathbb{R}^n}\quad & f(\mathbf{x}) \\
\text{subject to}\quad & g_i(\mathbf{x}) \le 0, \quad i = 1,\dots,m, \\
& h_j(\mathbf{x}) = 0, \quad j = 1,\dots,p.
\end{aligned}
$$

**Explanation:**

Constrained optimization restricts the search for a minimizer of $f$ to the [feasible set](../01_Optimization_Fundamentals/04_feasible_sets.ipynb) defined by inequality and equality constraints. Unlike the unconstrained case, a minimizer need not have zero gradient: the objective may still be decreasing when the boundary of the feasible set stops further progress. Optimality is then governed by the balance between the objective gradient and the active-constraint gradients, formalized by the [KKT conditions](../03_Convex_Optimization/46_karush_kuhn_tucker_conditions.ipynb).

**Assumptions:**
- $f, g_i, h_j$ are continuously differentiable.
- The feasible set is nonempty.

**Numerical Example:**

Minimize the distance-squared to the origin on a line:
$$
\min_{x_1,x_2}\; x_1^2 + x_2^2
\quad\text{subject to}\quad x_1 + x_2 = 1.
$$

Form the Lagrangian $L = x_1^2 + x_2^2 - \lambda(x_1 + x_2 - 1)$. Stationarity gives
$$
2x_1 = \lambda,\quad 2x_2 = \lambda \;\Rightarrow\; x_1 = x_2,
$$
and the constraint $x_1 + x_2 = 1$ yields
$$
\mathbf{x}^\star = \left(\tfrac12, \tfrac12\right),\qquad f(\mathbf{x}^\star) = \tfrac12.
$$

The unconstrained minimizer $(0,0)$ is infeasible; the constraint pushes the solution to the closest feasible point.

```python
import sympy as sp

x_1, x_2, lam = sp.symbols("x_1 x_2 lambda")
objective = x_1**2 + x_2**2
equality = x_1 + x_2 - 1
lagrangian = objective - lam*equality

stationarity = [sp.diff(lagrangian, variable) for variable in (x_1, x_2, lam)]
solution = sp.solve(stationarity, [x_1, x_2, lam], dict=True)[0]
minimizer = (solution[x_1], solution[x_2])
minimum_value = objective.subs(solution)

print("stationarity system =", stationarity)
print("minimizer =", minimizer)
print("minimum value =", minimum_value)
```

**Equivalent casadi implementation:**

```python
import casadi as ca

decision_variable = ca.SX.sym("x", 2)
objective = decision_variable[0]**2 + decision_variable[1]**2
equality = decision_variable[0] + decision_variable[1] - 1

solver = ca.nlpsol("solver", "ipopt", {"x": decision_variable, "f": objective, "g": equality},
                   {"ipopt.print_level": 0, "print_time": 0, "ipopt.sb": "yes"})
solution = solver(x0=[0, 0], lbg=0, ubg=0)

print("minimizer =", solution["x"])
print("minimum value =", float(solution["f"]))
```
