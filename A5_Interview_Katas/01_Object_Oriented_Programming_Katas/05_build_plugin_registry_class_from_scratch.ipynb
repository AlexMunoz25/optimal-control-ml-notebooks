### A5.1.5. Build Plugin Registry Class from Scratch

**Problem:**

Build a registry that maps string names to classes. Plugins register themselves by name, and clients look up and instantiate plugins by name at runtime.

**Explanation:**

A plugin registry is a dictionary that maps names to classes. Code registers a class under a name, and later code can retrieve and instantiate that class by name alone, without importing it directly.

How to build it:

1. Store a dictionary mapping names to classes in `__init__`.
2. `register(name, cls)`: store the class in the dictionary under the given name.
3. `create(name, *args, **kwargs)`: look up the class by name and call it with the provided arguments to create an instance.
4. Optionally, provide a `register` decorator that registers a class under its own name.

**Example:**

Register a `JsonParser` class under the name `"json"`. Later, call `registry.create("json", data)` to get a `JsonParser` instance without importing `JsonParser` directly.

```python
class PluginRegistry:
    def __init__(self):
        self.plugins = {}

    def register(self, name, cls):
        self.plugins[name] = cls

    def create(self, name, *args, **kwargs):
        return self.plugins[name](*args, **kwargs)


class JsonParser:
    def __init__(self, data):
        self.data = data

    def parse(self):
        return f"Parsing JSON: {self.data}"


class XmlParser:
    def __init__(self, data):
        self.data = data

    def parse(self):
        return f"Parsing XML: {self.data}"


registry = PluginRegistry()
registry.register("json", JsonParser)
registry.register("xml", XmlParser)

parser = registry.create("json", '{"key": "value"}')
print(parser.parse())

parser = registry.create("xml", "<root/>")
print(parser.parse())
```
