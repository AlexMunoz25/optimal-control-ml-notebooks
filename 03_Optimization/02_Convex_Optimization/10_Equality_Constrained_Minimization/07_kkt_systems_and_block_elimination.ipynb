### 3.2.10.7. KKT Systems and Block Elimination


$$\begin{bmatrix}H&A^T\\A&0\end{bmatrix}\begin{bmatrix}u\\v\end{bmatrix}=\begin{bmatrix}r\\s\end{bmatrix}$$


**Explanation:**

KKT block systems are the computational form of equality-constrained stationarity and feasibility equations. Feasibility is the equality constraint or its linearized version; optimality is stationarity together with that feasibility condition.

**Properties:**
- Equality constraints introduce unrestricted multipliers.
- KKT systems couple stationarity and feasibility.
- Eliminating constraints can reduce the number of variables.

**Intuition:**

The numerical example keeps the equality constraint visible in every algebraic step.


**Numerical Example:**

Solve
$$
\begin{bmatrix}4&1\\1&3\end{bmatrix}
\begin{bmatrix}u_1\\u_2\end{bmatrix}
=
\begin{bmatrix}1\\2\end{bmatrix}.
$$

The determinant is
$$
4\cdot3-1\cdot1=11.
$$

Thus
$$
u_1=\frac{3\cdot1-1\cdot2}{11}=\frac1{11},
\qquad
u_2=\frac{-1\cdot1+4\cdot2}{11}=\frac7{11}.
$$


```python
import sympy as sp

coefficient_matrix = sp.Matrix([[4, 1], [1, 3]])
right_side = sp.Matrix([1, 2])
solution = coefficient_matrix.LUsolve(right_side)

solution
```
