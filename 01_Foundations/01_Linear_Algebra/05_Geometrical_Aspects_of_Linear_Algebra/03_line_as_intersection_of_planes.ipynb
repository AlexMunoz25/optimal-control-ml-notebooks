### 1.1.5.3. Line as the Intersection of Two Planes


$$
\begin{aligned}
A_1 x + B_1 y + C_1 z &= D_1 \\
A_2 x + B_2 y + C_2 z &= D_2
\end{aligned}
$$


**Explanation:**

In Savov's geometric treatment, two nonparallel planes intersect in a line. This notebook isolates that single idea: the line is obtained by solving the two plane equations simultaneously and leaving one free parameter. It relies on the definition of a [plane](./02_plane_in_r3.ipynb) and links back to the computational workflow of [reduced row echelon form](../04_Computational_Linear_Algebra/03_reduced_row_echelon_form.ipynb).

**Example:**

If

$$
\begin{aligned}
z &= 0 \\
y + z &= 0,
\end{aligned}
$$

then the reduced system implies

$$
y = 0,
\qquad
z = 0,
\qquad
x = t,
$$

so the intersection line is

$$
\ell = \left\{
\begin{bmatrix}
0 \\
0 \\
0
\end{bmatrix}
+ t
\begin{bmatrix}
1 \\
0 \\
0
\end{bmatrix}
\mid t \in \mathbb{R}
\right\}.
$$


```python
import sympy as sp

coefficient_matrix = sp.Matrix([[0, 0, 1], [0, 1, 1]])
constant_vector = sp.Matrix([0, 0])
augmented_matrix = coefficient_matrix.row_join(constant_vector)
reduced_matrix, pivot_columns = augmented_matrix.rref()

print("reduced_matrix =")
sp.pprint(reduced_matrix)
print("pivot_columns =", pivot_columns)
```
