### 1.1.3.9. Cramer's Rule

$$
x_i = \frac{\det(\mathbf{A}_i)}{\det(\mathbf{A})},
\quad
\det(\mathbf{A}) \neq 0
$$

**Explanation:**

Cramer's Rule solves a system of linear equations $\mathbf{A}\mathbf{x} = \mathbf{b}$
by replacing each column of $\mathbf{A}$ with $\mathbf{b}$ and computing determinants.
It is a theoretical method often used for small systems.

Because Cramer's Rule replaces columns and compares determinants, it sits naturally beside the [determinant of a 2×2 matrix](./07_matrix_determinant_2d.ipynb) and the explicit formula for the [inverse of a 2×2 matrix](./08_matrix_inverse_2d.ipynb). It solves the same systems later written compactly as [matrix equations](../04_Computational_Linear_Algebra/06_matrix_equations.ipynb).

**Example:**

Solve

$$
\begin{bmatrix}2&1\\1&3\end{bmatrix}
\begin{bmatrix}x_1\\x_2\end{bmatrix}
=
\begin{bmatrix}5\\7\end{bmatrix}.
$$

The coefficient determinant is

$$
D=2\cdot3-1\cdot1=5.
$$

Replace the first column by the right-hand side:

$$
D_1=\det\begin{bmatrix}5&1\\7&3\end{bmatrix}=5\cdot3-1\cdot7=8.
$$

Replace the second column by the right-hand side:

$$
D_2=\det\begin{bmatrix}2&5\\1&7\end{bmatrix}=2\cdot7-5\cdot1=9.
$$

Therefore

$$
x_1=\frac{D_1}{D}=\frac{8}{5},
\qquad
x_2=\frac{D_2}{D}=\frac{9}{5}.
$$


```python
import sympy as sp

coefficient_matrix = sp.Matrix([[2, 1], [1, 3]])
constants = sp.Matrix([5, 7])

det_coeff = coefficient_matrix.det()

num_vars = coefficient_matrix.cols
replaced_matrices = [
    sp.Matrix.hstack(*[
        constants if col == idx else coefficient_matrix.col(col)
        for col in range(num_vars)
    ])
    for idx in range(num_vars)
]
solution = [replaced.det() / det_coeff for replaced in replaced_matrices]

print("A =")
sp.pprint(coefficient_matrix)
print("b =", list(constants))
print("det(A) =", det_coeff)
print("Solution x =", solution)
```
