### 3.2.10.3. Dual Solution of Equality Constrained Problems


$$g(\nu)=\inf_x(f_0(x)+\nu^T(Ax-b))$$


**Explanation:**

The equality-constrained dual minimizes the Lagrangian over primal variables and then maximizes the resulting concave dual function over the equality multiplier. Feasibility is the equality constraint or its linearized version; optimality is stationarity together with that feasibility condition.

**Properties:**
- Equality constraints introduce unrestricted multipliers.
- KKT systems couple stationarity and feasibility.
- Eliminating constraints can reduce the number of variables.

**Intuition:**

The numerical example keeps the equality constraint visible in every algebraic step.


**Numerical Example:**

For
$$
\min \frac12(x_1^2+x_2^2)\quad\text{subject to}\quad x_1+x_2=2,
$$
the Lagrangian is
$$
L=\frac12(x_1^2+x_2^2)+\nu(x_1+x_2-2).
$$

Stationarity gives
$$
x_1=-\nu,\qquad x_2=-\nu.
$$

Feasibility gives
$$
-\nu-\nu=2\quad\Longrightarrow\quad \nu=-1.
$$

Therefore $x_1=x_2=1$.


```python
import sympy as sp

variable_1, variable_2, multiplier = sp.symbols('x_1 x_2 nu')
solution = sp.solve([sp.Eq(variable_1 + multiplier, 0), sp.Eq(variable_2 + multiplier, 0), sp.Eq(variable_1 + variable_2, 2)], [variable_1, variable_2, multiplier])

solution
```
