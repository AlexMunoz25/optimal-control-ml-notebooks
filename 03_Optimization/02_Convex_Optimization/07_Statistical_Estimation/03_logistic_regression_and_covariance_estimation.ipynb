### 3.2.7.3. Logistic Regression and Covariance Estimation


$$\operatorname{prob}(y=1\mid u)=\frac{\exp(au+b)}{1+\exp(au+b)}$$


**Explanation:**

Logistic regression models binary probabilities using a sigmoid applied to an affine score, while covariance estimation uses matrix variables constrained by positive semidefiniteness. The logistic negative log-likelihood is convex in the affine parameters for fixed data. Feasibility is usually the parameter domain or matrix cone constraint; optimality minimizes the chosen likelihood objective.

The response $y$ is binary and $u$ is the scalar feature in this simplified display. The affine score $au+b$ is passed through the logistic function, producing a probability between zero and one.

**Properties:**
- The logistic probability lies between zero and one.
- The positive-label loss is log(1+exp(score))-score.
- Covariance matrices must be positive semidefinite.

**Intuition:**

The figure above shows a logistic fit; the calculation evaluates one score on the same sigmoid curve.

**Numerical Example:**

The figure above shows a logistic fit; this calculation evaluates one point on the same sigmoid curve.

Let the affine score be
$$
s=au+b,
$$
with $a=2$, $u=1$, and $b=-1$. Then
$$
s=2\cdot1-1=1.
$$

The probability of label $1$ is
$$
\frac{\exp(s)}{1+\exp(s)}=\frac{\exp(1)}{1+\exp(1)}.
$$

For an observed positive label, the negative log-likelihood contribution is
$$
-\log\left(\frac{\exp(1)}{1+\exp(1)}\right)
=\log(1+\exp(1))-1.
$$


```python
import sympy as sp

score = sp.Integer(1)
probability = sp.exp(score) / (1 + sp.exp(score))
positive_label_loss = sp.log(1 + sp.exp(score)) - score

probability, positive_label_loss
```
