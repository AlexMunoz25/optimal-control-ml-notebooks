### 3.1.20. Conjugate Direction Methods


$$
d_i^{\mathsf T}Qd_j=0\quad (i\ne j),
\qquad
x^{k+1}=x^k+\alpha^k d^k,
\qquad
\alpha^k=-\frac{\nabla f(x^k)^{\mathsf T}d^k}{d^{k\mathsf T}Qd^k}.
$$


**Explanation:**

Conjugate direction methods are designed for positive definite quadratic costs. Ordinary orthogonal directions are natural for a spherical quadratic, but an elliptical quadratic has a different geometry determined by $Q$. Directions are $Q$-conjugate when they are orthogonal under the inner product induced by $Q$, and exact minimization along $n$ such directions reaches the quadratic minimizer in at most $n$ steps. This generalizes the idea of [orthogonality](../../01_Foundations/01_Linear_Algebra/02_Vector/07_vector_orthogonality_condition.ipynb) to the geometry of the objective.

Bertsekas uses this topic to show how choosing better directions can eliminate the zig-zagging of steepest descent. The method is still based on line minimization, but each new direction is constructed so that progress made along previous directions is not undone. This is the finite-dimensional geometric heart of conjugate gradient.

**Example:**

The first figure explains conjugate directions by transforming an ordinary orthogonal-coordinate minimization into the coordinates of a quadratic. The second shows a [Gram-Schmidt orthogonalization](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/07_gram_schmidt_orthogonalization.ipynb)-like construction, but with the $Q$-inner product replacing the Euclidean one. The code cell builds a second direction that is conjugate to the first for a small positive definite matrix.

```python
import numpy as np

Q = np.array([[4.0, 1.0], [1.0, 3.0]])
first_direction = np.array([1.0, 0.0])
trial_direction = np.array([0.0, 1.0])
second_direction = trial_direction - (trial_direction @ Q @ first_direction) / (first_direction @ Q @ first_direction) * first_direction
conjugacy_check = first_direction @ Q @ second_direction

second_direction, conjugacy_check
```
