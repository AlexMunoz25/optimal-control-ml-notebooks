### 3.1.16. Least Squares Problem Models


$$
\min_{x\in\mathbb{R}^n} f(x),
\qquad
f(x)=\frac{1}{2}\|g(x)\|^2
=\frac{1}{2}\sum_{i=1}^m g_i(x)^2.
$$


**Explanation:**

Least squares problems measure the size of a residual vector. They appear whenever a model is chosen so that predicted outputs match data, measurements, or desired behavior as closely as possible. Curve fitting, dynamic system identification, neural networks, and pattern classification are different instances of the same mathematical form. The residual norm is a [vector norm](../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb), and in the linear case the problem connects directly to projections and normal equations.

The decision vector $x$ is chosen to minimize the squared [norm](../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb) of a residual vector $g(x)$. The equality with the sum shows that the objective adds the squared residual components $g_i(x)^2$ and multiplies the result by $1/2$ for cleaner derivatives.

For students new to optimization, least squares is a useful bridge between linear algebra and nonlinear programming. The familiar linear least squares problem becomes nonlinear when the residuals depend nonlinearly on the parameters. The objective still has the same sum-of-squares structure, but the optimization landscape can now be curved or nonconvex.

**Intuition:**

The figure shows a least squares cost surface for a neural-network example, emphasizing that even a sum of squares can have a nontrivial shape. The code cell solves a small linear fitting problem so that the basic residual-minimization idea is visible before moving to nonlinear Gauss-Newton steps.

**Numerical Example:**

The least-squares surface figure is generated by residual sizes. Fit a line $y=at+b$ to
$$
(t,y)=(0,1),\ (1,2),\ (2,2.9),\ (3,4.2).
$$

The design matrix and observation vector are
$$
A=\begin{bmatrix}0&1\\1&1\\2&1\\3&1\end{bmatrix},
\qquad y=\begin{bmatrix}1\\2\\2.9\\4.2\end{bmatrix}.
$$

The normal equations are
$$
A^{\mathsf T}A\begin{bmatrix}a\\b\end{bmatrix}=A^{\mathsf T}y,
\qquad
\begin{bmatrix}14&6\\6&4\end{bmatrix}
\begin{bmatrix}a\\b\end{bmatrix}
=
\begin{bmatrix}20.4\\10.2\end{bmatrix}.
$$

From $6a+4b=10.2$, $b=2.55-1.5a$. Substitute into the first equation:
$$
14a+6(2.55-1.5a)=20.4
\quad\Longrightarrow\quad
5a=5.1
\quad\Longrightarrow\quad
a=1.02,
\qquad b=1.02.
$$

The fitted predictions are $(1.02,2.04,3.06,4.08)$, so the residuals are $(0.02,0.04,0.16,-0.12)$ and the residual sum of squares is $0.042$.


```python
import numpy as np

sample_points = np.array([0.0, 1.0, 2.0, 3.0])
observations = np.array([1.0, 2.0, 2.9, 4.2])
design_matrix = np.column_stack([sample_points, np.ones_like(sample_points)])
parameters, residuals, rank, singular_values = np.linalg.lstsq(design_matrix, observations, rcond=None)

print("parameters =", parameters)
print("residuals =", residuals)
```
