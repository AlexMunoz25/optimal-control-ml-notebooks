### 1.1.6.8. Affine Transformation


$$
A(\mathbf{x}) = M_T \mathbf{x} + \mathbf{b}
$$


**Explanation:**

An affine transformation has the form of a linear transformation plus a translation, so it does not preserve the origin. This is the key distinction from a [linear transformation](./01_linear_transformation.ipynb). Affine maps are common in geometry and graphics even though they are not linear in the strict algebraic sense.

**Example:**

If

$$
M_T =
\begin{bmatrix}
2 & 0 \\
0 & 1
\end{bmatrix},
\qquad
\mathbf{b} =
\begin{bmatrix}
1 \\
-2
\end{bmatrix},
$$

then

$$
A
\left(
\begin{bmatrix}
3 \\
5
\end{bmatrix}
\right)
=
\begin{bmatrix}
7 \\
3
\end{bmatrix}.
$$


```python
import sympy as sp

linear_part = sp.Matrix([[2, 0], [0, 1]])
translation_part = sp.Matrix([1, -2])
input_vector = sp.Matrix([3, 5])
output_vector = linear_part * input_vector + translation_part

print("output_vector =", output_vector)
```
