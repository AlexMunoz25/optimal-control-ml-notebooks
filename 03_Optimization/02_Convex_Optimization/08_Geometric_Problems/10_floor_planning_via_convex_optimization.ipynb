### 3.2.8.10. Floor Planning via Convex Optimization


$$x_i+w_i+\rho\leq x_j$$


**Explanation:**

Floor-planning models express relative placement, spacing, containment, and shape requirements as optimization constraints. Many placement constraints are affine, while area and aspect-ratio constraints require convex reformulation. Feasibility means every layout inequality is satisfied.

The variables $x_i$ and $x_j$ are horizontal positions of two blocks, $w_i$ is the width of block $i$, and $\rho$ is required spacing. The inequality says block $i$ must end at least $\rho$ before block $j$ begins.

**Properties:**
- Relative positioning constraints are linear.
- Spacing constraints add margins between blocks.
- Geometric programming handles some area and aspect-ratio constraints.

**Intuition:**

The example checks one horizontal spacing constraint exactly.


**Numerical Example:**

Suppose block $i$ starts at $x_i=0$, has width $w_i=2$, and must leave spacing $\rho=1$ before block $j$, whose start is $x_j=4$.

The constraint is
$$
x_i+w_i+\rho\leq x_j.
$$

Substitute values:
$$
0+2+1=3\leq4.
$$

The placement satisfies the spacing constraint.


```python
import sympy as sp

block_i_position = sp.Integer(0)
block_i_width = sp.Integer(2)
spacing = sp.Integer(1)
block_j_position = sp.Integer(4)
left_side = block_i_position + block_i_width + spacing

print("left_side =", left_side)
print("left_side <= block_j_position =", left_side <= block_j_position)
```
