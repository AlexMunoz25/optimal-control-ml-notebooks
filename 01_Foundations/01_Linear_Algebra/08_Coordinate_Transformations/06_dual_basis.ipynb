### 1.1.8.6. Dual Basis


$$
\mathbf{e}^{\,i} \cdot \mathbf{e}_j = \delta^i_j
$$


**Explanation:**

Fleisch introduces the dual basis to recover a clean coefficient-extraction rule in non-orthogonal coordinates. Each dual vector is orthogonal to all primal basis vectors with different indices and normalized against the matching basis vector. This notebook depends on the notebook on the [non-orthogonal coordinate system](./05_non_orthogonal_coordinate_system.ipynb).

**Example:**

Once the dual basis is known, coefficients can be extracted by dotting with the dual vectors instead of solving a new linear system each time.


```python
import sympy as sp

basis_matrix = sp.Matrix([[1, sp.Rational(1, 2)], [0, sp.sqrt(3) / 2]])
dual_basis_matrix = sp.simplify(basis_matrix.inv().T)

print("dual_basis_matrix =")
sp.pprint(dual_basis_matrix)
print("biorthogonality_check =")
sp.pprint(sp.simplify(dual_basis_matrix.T * basis_matrix))
```
