### 3.9.28. Incremental Gradient Methods

$$
f(x)=\sum_{i=1}^m f_i(x),
\qquad
x_{i+1}^k = x_i^k - \alpha^k \nabla f_i(x_i^k),
\qquad
x^{k+1}=x_m^k.
$$


**Explanation:**

Incremental gradient methods exploit objectives that are sums of many component functions. Instead of computing the full gradient of $f=\sum_i f_i$ before every update, the method processes one component or data block at a time. This can reduce the cost of each update and can produce useful progress before a full pass through all data has been completed. These methods fit naturally with least-squares problems because large data-fitting problems naturally decompose into residual blocks.

The objective is written as a sum of component costs $f_i$. Instead of using the full gradient at once, the inner update moves through component gradients one at a time, and the final inner point becomes the next outer iterate.

The price of incrementalism is that the direction at an intermediate substep is not the true full gradient direction. Far from the solution, this can be beneficial because each component quickly pulls the iterate toward satisfying one part of the model. Near the solution, however, the analysis becomes more delicate because component gradients may not vanish simultaneously at every substep.

**Intuition:**

For scalar linear fitting, each data point defines one residual term and one component gradient. A cycle of incremental updates visits these terms one at a time, adjusting the parameter after each residual. The code cell shows one such pass through four data points, illustrating the mechanics before any convergence theorem is invoked.


**Numerical Example:**

For scalar fitting with residual $r_i(w)=wa_i-y_i$, use
$$
(a_i,y_i)=(1,1.1),(2,2.1),(3,2.9),(4,4.2),
\qquad w_0=0,
\qquad \alpha=0.05.
$$

The component gradient is $r_i(w)a_i$. Process the data one point at a time:
$$
\begin{aligned}
r_1&=0\cdot 1-1.1=-1.1,
& w_1&=0-0.05(-1.1)(1)=0.055,\\
r_2&=0.055\cdot 2-2.1=-1.99,
& w_2&=0.055-0.05(-1.99)(2)=0.254,\\
r_3&=0.254\cdot 3-2.9=-2.138,
& w_3&=0.254-0.05(-2.138)(3)=0.5747,\\
r_4&=0.5747\cdot 4-4.2=-1.9012,
& w_4&=0.5747-0.05(-1.9012)(4)=0.95494.
\end{aligned}
$$

After one incremental pass, the parameter has moved from $0$ to approximately $0.955$.


```python
import sympy as sp

p, a, y = sp.symbols("p a y")
sample_loss = (p * a - y) ** 2 / 2
sample_gradient = sp.diff(sample_loss, p)

features = [1, 2, 3, 4]
observations = [sp.Rational(11, 10), sp.Rational(21, 10), sp.Rational(29, 10), sp.Rational(21, 5)]
parameter = sp.Integer(0)
stepsize = sp.Rational(1, 20)

for feature, observation in zip(features, observations):
    gradient_value = sample_gradient.subs({p: parameter, a: feature, y: observation})
    parameter = parameter - stepsize * gradient_value

print("sample_gradient =", sample_gradient)
print("parameter =", parameter, "=", float(parameter))
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

parameter_symbol = ca.SX.sym("p")
feature_symbol = ca.SX.sym("a")
observation_symbol = ca.SX.sym("y")
sample_loss = 0.5 * (parameter_symbol * feature_symbol - observation_symbol) ** 2
sample_gradient = ca.gradient(sample_loss, parameter_symbol)
sample_gradient_function = ca.Function("g", [parameter_symbol, feature_symbol, observation_symbol], [sample_gradient])

features = [1.0, 2.0, 3.0, 4.0]
observations = [1.1, 2.1, 2.9, 4.2]
parameter = 0.0
stepsize = 0.05

for sample_index in range(len(features)):
    feature = features[sample_index]
    observation = observations[sample_index]
    gradient_value = float(sample_gradient_function(parameter, feature, observation))
    parameter = parameter - stepsize * gradient_value
    print(f"sample {sample_index + 1}: gradient = {gradient_value:.4f}, parameter = {parameter:.4f}")

print("final_parameter =", parameter)
```
