### 1.1.8.7. Contravariant Components


$$
\mathbf{A} = A^i \mathbf{e}_i,
\qquad
A^i = \mathbf{A} \cdot \mathbf{e}^{\,i},
\qquad
A^{i'} = \frac{\partial x^{i'}}{\partial x^j} A^j
$$


**Explanation:**

Contravariant components are the coefficients in the expansion of a vector with respect to the original basis vectors. Geometrically, they correspond to the parallel-projection construction in a skew basis, and algebraically they are extracted by dotting with the dual basis vectors. Because the basis vectors transform directly, these upper-index components transform with the inverse law so that the vector itself remains unchanged.

</p>

**Example:**

For
$$
\mathbf{A} =
\begin{bmatrix}
7 \\
2
\end{bmatrix},
\qquad
\mathbf{e}_1 =
\begin{bmatrix}
1 \\
3
\end{bmatrix},
\qquad
\mathbf{e}_2 =
\begin{bmatrix}
4 \\
0
\end{bmatrix},
$$
the contravariant components are
$$
A^1 = \mathbf{A} \cdot \mathbf{e}^{\,1} = \frac{2}{3},
\qquad
A^2 = \mathbf{A} \cdot \mathbf{e}^{\,2} = \frac{19}{12}.
$$


```python
import sympy as sp

basis_vector_1 = sp.Matrix([1, 3])
basis_vector_2 = sp.Matrix([4, 0])
vector_a = sp.Matrix([7, 2])
primal_basis_matrix = sp.Matrix.hstack(basis_vector_1, basis_vector_2)
dual_basis_matrix = primal_basis_matrix.inv().T
contravariant_components_from_linear_system = primal_basis_matrix.LUsolve(vector_a)
contravariant_components_from_dual_basis = sp.Matrix(
    [vector_a.dot(dual_basis_matrix[:, 0]), vector_a.dot(dual_basis_matrix[:, 1])]
)

print("contravariant_components_from_linear_system =", contravariant_components_from_linear_system)
print("contravariant_components_from_dual_basis =", contravariant_components_from_dual_basis)
```
