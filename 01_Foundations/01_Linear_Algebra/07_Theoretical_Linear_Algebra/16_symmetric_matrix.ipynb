### 1.1.7.16. Symmetric Matrix


$$
A^{\mathsf{T}} = A
$$


**Explanation:**

A symmetric matrix equals its transpose. Its main structural consequences are real eigenvalues and orthogonal eigenvectors, while the stronger positivity condition is treated separately in [positive definite matrices](./10_positive_definite_matrix.ipynb).

**Example:**

For

$$
A=\begin{bmatrix}2&1\\1&3\end{bmatrix},
$$

transposing swaps the off-diagonal entries, but both off-diagonal entries are $1$:

$$
A^T=\begin{bmatrix}2&1\\1&3\end{bmatrix}=A.
$$

The characteristic polynomial is

$$
\det(A-\lambda I)=(2-\lambda)(3-\lambda)-1=\lambda^2-5\lambda+5.
$$

Thus the eigenvalues are

$$
\lambda=\frac{5\pm\sqrt{5}}{2}.
$$


```python
import sympy as sp

matrix_a = sp.Matrix([[2, 1], [1, 3]])

print("is_symmetric =", matrix_a == matrix_a.T)
print("eigenvalues =", list(matrix_a.eigenvals().keys()))
```
