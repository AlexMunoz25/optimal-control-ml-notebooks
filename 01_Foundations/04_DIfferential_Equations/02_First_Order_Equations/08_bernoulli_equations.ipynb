### 1.4.2.8. Bernoulli Equations

$$
\frac{dy}{dx} + P(x)\,y = f(x)\,y^{n}
\;\xrightarrow{\,u = y^{1-n}\,}\;
\frac{du}{dx} + (1 - n)P(x)\,u = (1 - n)f(x) \ \text{(linear)} .
$$

**Explanation:**

**Bernoulli's equation** has the form $y' + P(x)y = f(x)y^n$. For $n = 0$ or $n = 1$ it is already [linear](./04_linear_first_order_equations.ipynb); for any other $n$ the substitution $u = y^{1-n}$ turns it into a **linear** equation in $u$, which is solved by an integrating factor and then back-substituted. It is the standard way to linearize an important class of nonlinear first-order equations — including the logistic equation written in this form.

**Properties:**
- The substitution works because $u = y^{1-n}$ gives $u' = (1-n)y^{-n}y'$, which clears the $y^n$ term.
- $y = 0$ may be a singular solution lost in the division by $y^n$; check it separately.

**Numerical Example:**

Solve $x\dfrac{dy}{dx} + y = x^2 y^2$.

**Step 1 — standard form.** Divide by $x$:

$$
\frac{dy}{dx} + \frac{1}{x}\,y = x\,y^2, \qquad \text{so } P = \tfrac1x,\ f = x,\ n = 2 .
$$

**Step 2 — substitute** $u = y^{1-2} = y^{-1}$, i.e. $y = u^{-1}$, so by the chain rule $\dfrac{dy}{dx} = -u^{-2}\dfrac{du}{dx}$. Inserting and multiplying through by $-u^2$:

$$
-u^{-2}\frac{du}{dx} + \frac{1}{x}u^{-1} = x\,u^{-2}
\;\Longrightarrow\; \frac{du}{dx} - \frac{1}{x}\,u = -x .
$$

**Step 3 — solve the linear equation.** The integrating factor is $\mu = e^{-\int dx/x} = x^{-1}$, so

$$
\frac{d}{dx}\big[x^{-1}u\big] = -1 \;\Rightarrow\; x^{-1}u = -x + c \;\Rightarrow\; u = -x^2 + cx .
$$

**Step 4 — back-substitute** $u = y^{-1}$:

$$
y = \frac{1}{u} = \frac{1}{-x^2 + cx}.
$$

```python
import sympy as sp

x = sp.symbols("x")
y = sp.Function("y")

bernoulli = sp.Eq(y(x).diff(x) + y(x) / x, x * y(x)**2)
solution = sp.dsolve(bernoulli, y(x), hint="Bernoulli")

verification = sp.simplify(bernoulli.lhs.subs(y(x), solution.rhs).doit() - bernoulli.rhs.subs(y(x), solution.rhs))

print("Bernoulli equation : y' + y/x = x y^2  (n = 2)")
print("substitution u = y^(1-n) = y^(-1) gives the linear equation u' - u/x = -x")
print("solution           :", solution)
print("check (LHS - RHS)  :", verification)
```
