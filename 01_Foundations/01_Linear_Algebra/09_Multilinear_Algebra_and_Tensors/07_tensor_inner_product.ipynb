### 1.1.9.7. Tensor Inner Product

$$
A^i B_i,
\qquad
C^i{}_k = A^{ij} B_{jk},
\qquad
\operatorname{rank}(C) = \operatorname{rank}(A) + \operatorname{rank}(B) - 2
$$

**Explanation:**

The inner product is formed by taking an outer product and then contracting one upper-lower index pair. The dot product from [Vector Dot Product](../02_Vector/04_vector_dot_product.ipynb) and the product from [Matrix Multiplication](../03_Matrix/05_matrix_multiplication.ipynb) are the familiar low-rank cases. Every contraction lowers the total rank by two, so the result has lower rank than the full outer product.

**Example:**

For vectors

$$
\mathbf{a}=\begin{bmatrix}1\\2\\3\end{bmatrix},
\qquad
\mathbf{b}=\begin{bmatrix}4\\5\\6\end{bmatrix},
$$

the tensor inner product contracts the outer product:

$$
\mathbf{a}\cdot\mathbf{b}=1\cdot4+2\cdot5+3\cdot6=4+10+18=32.
$$

For matrices

$$
A=\begin{bmatrix}1&2\\3&4\end{bmatrix},
\qquad
B=\begin{bmatrix}5&6\\7&8\end{bmatrix},
$$

the contraction used here matches matrix multiplication:

$$
AB=\begin{bmatrix}19&22\\43&50\end{bmatrix}.
$$


```python
import sympy as sp

vector_a = sp.Array([1, 2, 3])
vector_b = sp.Array([4, 5, 6])
matrix_a = sp.Array([[1, 2], [3, 4]])
matrix_b = sp.Array([[5, 6], [7, 8]])

vector_inner_product = sp.tensorcontraction(sp.tensorproduct(vector_a, vector_b), (0, 1))
matrix_inner_product = sp.tensorcontraction(sp.tensorproduct(matrix_a, matrix_b), (1, 2))

print("vector_inner_product =", vector_inner_product)
print("matrix_inner_product =")
sp.pprint(matrix_inner_product)
print("matrix_a * matrix_b =")
sp.pprint(sp.Matrix(matrix_a) * sp.Matrix(matrix_b))
```
