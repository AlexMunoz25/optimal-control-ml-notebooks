### 3.1.5. Quadratic Cost Functions and Existence


$$
\begin{aligned}
f(x) &= \frac{1}{2}x^{\mathsf T}Qx - b^{\mathsf T}x + c, \\
\nabla f(x) &= Qx-b, \\
Q \succ 0 &\Longrightarrow x^* = Q^{-1}b\ \text{is the unique global minimum}.
\end{aligned}
$$


**Explanation:**

Quadratic cost functions are the simplest nonlinear models that still contain the main geometry of unconstrained optimization. Their gradients are affine, their Hessians are constant, and the definiteness of the matrix $Q$ largely determines the behavior of the problem. When $Q$ is positive definite, the stationary equation $Qx=b$ has the unique solution $x^*=Q^{-1}b$, using the [matrix inverse](../../01_Foundations/01_Linear_Algebra/03_Matrix/11_matrix_inverse.ipynb), and this point is the unique global minimum. When $Q$ is singular or indefinite, existence and uniqueness require more care, and the [eigenvalues](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/08_eigenvalue.ipynb) of $Q$ reveal what can go wrong.

Bertsekas gives special attention to quadratics because later convergence-rate analysis repeatedly reduces smooth problems to local quadratic approximations. In that sense, a quadratic is not merely an example; it is the model problem behind gradient, Newton, conjugate-gradient, and quasi-Newton behavior.

The first figure displays the main quadratic cases through level sets, including the positive definite case with elliptical contours and the cases where minimizers fail to exist or fail to be unique. The code cell solves one positive definite quadratic by solving its stationary linear system.

```python
import numpy as np

Q = np.array([[4.0, 1.0], [1.0, 2.0]])
linear_term = np.array([1.0, 3.0])
minimizer = np.linalg.solve(Q, linear_term)
minimum_value = 0.5 * minimizer @ Q @ minimizer - linear_term @ minimizer

minimizer, minimum_value
```
