### 3.2.5.9. Theorems of Alternatives


$$\text{exactly one associated feasibility system holds}$$


**Explanation:**

Theorems of alternatives express infeasibility through certificates. For a pair of alternative systems, exactly one can hold: either the original feasibility statement has a solution, or a separating certificate proves that it cannot.

The displayed statement is a logical form, not a numeric formula. It says that for a paired set of feasibility systems, one and only one system can be feasible, so infeasibility of one side is certified by feasibility of the other.

**Properties:**
- Weak duality gives a lower bound for minimization.
- A primal feasible point gives an upper bound.
- A zero gap proves optimality when both sides are feasible.

**Intuition:**

The example treats the numbers as certificates rather than as iterates.


**Numerical Example:**

Suppose a primal feasible point has value $p=4$ and a dual feasible point has value $d=3$. Weak duality gives
$$
d\leq p^\star\leq p.
$$

Substitute the values:
$$
3\leq p^\star\leq4.
$$

The certificate gap is
$$
p-d=4-3=1.
$$

The dual point proves that no feasible primal solution can have value below $3$, while the primal point proves the optimum is at most $4$.


```python
import sympy as sp

primal_value = sp.Integer(4)
bound_value = sp.Integer(3)
gap = primal_value - bound_value

print("bound_value <= primal_value =", bound_value <= primal_value)
print("gap =", gap)
```
