### 1.3.7.1. Derivatives with Respect to a Vector

$$
\nabla_{\mathbf{x}} f = \begin{bmatrix} \partial f/\partial x_1 \\ \vdots \\ \partial f/\partial x_n \end{bmatrix}
\in \mathbb{R}^{n},
\qquad
\frac{\partial \mathbf{y}}{\partial \mathbf{x}} =
\begin{bmatrix} \partial y_i/\partial x_j \end{bmatrix} \in \mathbb{R}^{m\times n}\ (\text{Jacobian}) .
$$

**Explanation:**

Differentiating with respect to a vector collects partial derivatives into structured arrays. A scalar function gives the [gradient](../05_Multivariable_Differential_Calculus/03_gradient.ipynb) (a column vector), while a vector-valued map gives the [Jacobian](../05_Multivariable_Differential_Calculus/07_jacobian.ipynb) (a matrix whose rows are the gradients of the components). The only thing to fix is a *layout convention*: this track uses the **denominator layout**, where $\nabla_\mathbf{x} f$ is a column and $\partial\mathbf{y}/\partial\mathbf{x}$ has shape $m\times n$. Consistent layout is what makes the matrix-calculus identities and the AD rules that follow compose without transpose bookkeeping errors.

**Properties:**
- Linear map: $\dfrac{\partial (A\mathbf{x})}{\partial \mathbf{x}} = A$.
- For a scalar $f$, the gradient is the transpose of the $1\times n$ Jacobian: $\nabla_\mathbf{x} f = \big(\partial f/\partial\mathbf{x}\big)^\top$.

**Numerical Example:**

Let $\mathbf{y} = A\mathbf{x}$ with $A = \begin{bmatrix} 2 & 1 \\ 0 & 3 \end{bmatrix}$. Each output is $y_1 = 2x_1 + x_2$, $y_2 = 3x_2$, so

$$
\frac{\partial \mathbf{y}}{\partial \mathbf{x}} =
\begin{bmatrix} \partial y_1/\partial x_1 & \partial y_1/\partial x_2 \\ \partial y_2/\partial x_1 & \partial y_2/\partial x_2 \end{bmatrix}
= \begin{bmatrix} 2 & 1 \\ 0 & 3 \end{bmatrix} = A .
$$

For the scalar $f = \mathbf{a}^\top\mathbf{x}$ with $\mathbf{a} = [2, 3]^\top$, the gradient is $\nabla_\mathbf{x} f = \mathbf{a} = [2, 3]^\top$.

```python
import sympy as sp

x_1, x_2 = sp.symbols("x_1 x_2")
x = sp.Matrix([x_1, x_2])
A = sp.Matrix([[2, 1], [0, 3]])
a = sp.Matrix([2, 3])

vector_map = A * x
jacobian = vector_map.jacobian(x)

linear_form = (a.T * x)[0]
gradient = sp.Matrix([sp.diff(linear_form, variable) for variable in x])

print("∂(Ax)/∂x ="); sp.pprint(jacobian)
print("∇(aᵀx)   =", gradient.T, " (equals a)")
```

**Equivalent CasADi implementation:**

`ca.jacobian` and `ca.gradient` return the same Jacobian and gradient from the computational graph.

```python
import casadi as ca

x = ca.SX.sym("x", 2)
A = ca.DM([[2, 1], [0, 3]])
a = ca.DM([2, 3])

jacobian = ca.jacobian(A @ x, x)
gradient = ca.gradient(ca.dot(a, x), x)

print("∂(Ax)/∂x =", jacobian)
print("∇(aᵀx)   =", gradient)
```
