### 3.3.15. Regularized Approximation and Tradeoffs


$$\min_x\|Ax-b\|_2^2+\lambda\|x\|_2^2$$


**Explanation:**

Regularization adds a penalty to a fitting objective. The residual term pulls the solution toward the data, while the regularizer pulls it toward a simpler or smaller solution. Feasibility is unconstrained in this scalar example, and optimality is where the derivative of the combined objective vanishes.

The first term is the squared residual [norm](../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb), and the second term penalizes the size of the decision vector. The scalar $\lambda$ weights the tradeoff between fitting $Ax\approx b$ and keeping $x$ small.

**Properties:**
- Quadratic regularization preserves convexity.
- The regularization weight controls the tradeoff.
- The optimizer usually shifts away from the unregularized fit.

**Intuition:**

The computation solves the combined quadratic exactly.


**Numerical Example:**

Minimize
$$
f(x)=(x-2)^2+3x^2.
$$

Expand:
$$
f(x)=x^2-4x+4+3x^2=4x^2-4x+4.
$$

Differentiate:
$$
f'(x)=8x-4.
$$

Set the derivative to zero:
$$
8x-4=0 \quad\Longrightarrow\quad x=\frac12.
$$

The regularizer pulls the solution toward zero.


```python
import sympy as sp

variable = sp.symbols('x')
objective = (variable - 2) ** 2 + 3 * variable ** 2
optimizer = sp.solve(sp.diff(objective, variable), variable)[0]

print("optimizer =", optimizer)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

variable = ca.SX.sym("x")
objective = (variable - 2) ** 2 + 3 * variable ** 2
gradient = ca.gradient(objective, variable)
hessian = ca.jacobian(gradient, variable)
optimizer = ca.solve(ca.evalf(hessian), -ca.evalf(ca.substitute(gradient, variable, ca.DM(0))))

print("optimizer =", optimizer)
```
