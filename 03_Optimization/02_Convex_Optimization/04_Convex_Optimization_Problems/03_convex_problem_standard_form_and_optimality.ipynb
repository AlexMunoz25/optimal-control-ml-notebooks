### 3.2.4.3. Convex Problem Standard Form and Optimality


$$\nabla f_0(x^*)^T(y-x^*)\geq0\quad\forall y\ \text{feasible}$$


**Explanation:**

In a convex problem, standard form makes the global optimality certificate explicit. A differentiable objective is optimal over the feasible set when its first-order inequality rules out every feasible descent direction.

**Properties:**
- The gradient defines the tangent approximation.
- Convexity keeps the true function above that approximation.
- At an optimum, the first-order inequality holds against every feasible point.

**Intuition:**

The computation compares a tangent lower bound with the true value.


**Numerical Example:**

Let $f(x)=x^2$, expand around $x=1$, and compare with $y=3$. Since $f'(x)=2x$,
$$
f'(1)=2.
$$

The tangent lower bound at $y=3$ is
$$
f(1)+f'(1)(3-1)=1^2+2\cdot2=5.
$$

The true value is
$$
f(3)=3^2=9.
$$

Because $9\geq5$, the tangent line gives a valid global lower bound.


```python
import sympy as sp

point_x = sp.Integer(1)
point_y = sp.Integer(3)
lower_bound = point_x ** 2 + 2 * point_x * (point_y - point_x)
true_value = point_y ** 2

lower_bound, true_value
```
