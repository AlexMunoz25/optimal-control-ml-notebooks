### 1.1.6.10. Reflection Matrix Representation


$$
R_{\ell}(\mathbf{v}) = 2 \Pi_{\ell}(\mathbf{v}) - \mathbf{v}
$$


**Explanation:**

Savov derives reflections from projections: double the invariant component and subtract the original vector once. The geometric reflection itself depends on the earlier notebook on [projection onto a line](../05_Geometrical_Aspects_of_Linear_Algebra/07_projection_onto_a_line.ipynb), so this notebook concentrates on how the reflection becomes a matrix. Reflection is separated from rotation because the invariant structure is different.

**Example:**

Reflection across the $x$-axis uses

$$
M_{R_x} =
\begin{bmatrix}
1 & 0 \\
0 & -1
\end{bmatrix}.
$$


```python
import sympy as sp

reflection_matrix = sp.Matrix([[1, 0], [0, -1]])
input_vector = sp.Matrix([2, 3])
output_vector = reflection_matrix * input_vector

print("output_vector =", output_vector)
```
