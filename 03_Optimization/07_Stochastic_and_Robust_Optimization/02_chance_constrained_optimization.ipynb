### 3.7.2. Chance-Constrained Optimization

$$
\min_{\mathbf{x}}\; \mathbf{c}^\top \mathbf{x}
\quad\text{subject to}\quad
\Pr\big[\, g(\mathbf{x},\boldsymbol\xi) \le 0\,\big] \ge 1 - \varepsilon.
$$

**Explanation:**

A chance constraint requires a constraint to hold with at least probability $1-\varepsilon$ rather than for every realization of the uncertainty. When the uncertain quantity is Gaussian and enters linearly, the probabilistic constraint has an exact *deterministic equivalent*: a margin proportional to the standard deviation and the standard-normal quantile $\Phi^{-1}(1-\varepsilon)$. This converts the constraint into a [second-order cone](../04_Convex_Problem_Classes/04_second_order_cone_programming.ipynb) constraint, so the problem stays convex.

**Intuition:**

The stocking level is pushed out into the tail of the demand density until only an $\varepsilon$-fraction of probability mass violates the constraint.

![Chance constraint on a Gaussian demand](../../Figures/030702_chance_constraint_density.png)

**Numerical Example:**

Choose the smallest stock level $s$ that meets a Gaussian demand $D\sim N(100, 15^2)$ with probability at least $95\%$:
$$
\min_s\; s
\quad\text{s.t.}\quad
\Pr[D \le s] \ge 0.95.
$$

Standardizing, $\Pr[D\le s] = \Phi\!\left(\tfrac{s-100}{15}\right) \ge 0.95$ requires $\tfrac{s-100}{15} \ge \Phi^{-1}(0.95) = 1.645$, so
$$
s^\star = 100 + 1.645\cdot 15 = 124.7.
$$
The $24.7$-unit buffer above the mean demand is the price of the $95\%$ service level.

```python
import sympy as sp
from sympy.stats import Normal, cdf

mean, standard_deviation, epsilon = 100, 15, sp.Rational(5, 100)
demand = Normal("D", mean, standard_deviation)
standard_quantile = sp.nsolve(cdf(Normal("Z", 0, 1))(sp.Symbol("z")) - (1 - epsilon), sp.Symbol("z"), 1.6)
stock_level = mean + standard_quantile * standard_deviation

print("standard-normal 95% quantile =", round(float(standard_quantile), 3))
print("required stock level s =", round(float(stock_level), 1))
```

**Equivalent scipy.stats implementation:**

```python
from scipy.stats import norm

mean, standard_deviation, epsilon = 100, 15, 0.05
standard_quantile = norm.ppf(1 - epsilon)
stock_level = mean + standard_quantile * standard_deviation

print("standard-normal 95% quantile =", round(standard_quantile, 3))
print("required stock level s =", round(stock_level, 1))
```
