### 3.3.16. Linear Constraints and Lagrangian Duality

$$
q(\lambda,\mu)=\inf_{x\in X}\left(f(x)+\sum_{i=1}^{m}\lambda_i(e_i^T x-d_i)+\sum_{j=1}^{r}\mu_j(a_j^T x-b_j)\right)
$$

$$
\max_{\lambda,\mu} q(\lambda,\mu) \quad \text{subject to} \quad \mu\ge 0
$$

**Explanation:**

For convex cost functions with linear constraints, Lagrange multipliers exist under broad conditions and lead directly to a dual problem. The dual function is obtained by minimizing the Lagrangian over the simple set $X$, producing a lower bound on every primal feasible cost. Under the chapter's linear-constraint hypotheses, the best such lower bound equals the primal optimum, and optimal dual variables are Lagrange multipliers.

**Intuition:**

<img src="../../Figures/03030316_berts_fig_3_4_1_dual_function.jpeg" alt="Bertsekas Figure 3.4.1: dual function as a supporting line intercept and its connection to the primal function" style="width: 60%; height: auto;">

The figure shows the dual function as the intercept of a supporting line. The slope is set by the multiplier, and maximizing the intercept chooses the supporting line that touches the optimal value at the original constraint level.

**Numerical Example:**

Consider the scalar equality-constrained problem

$$
\min \frac{1}{2}(x-2)^2
\quad \text{subject to} \quad x=0.
$$

The Lagrangian is

$$
L(x,\lambda)=\frac{1}{2}(x-2)^2+\lambda x.
$$

The dual function minimizes this expression over $x$:

$$
\frac{\partial L}{\partial x}=x-2+\lambda=0
\quad \Rightarrow \quad
x_\lambda=2-\lambda.
$$

Substitute into the Lagrangian:

$$
q(\lambda)=\frac{1}{2}(-\lambda)^2+\lambda(2-\lambda)
=2\lambda-\frac{1}{2}\lambda^2.
$$

Maximize $q$:

$$
q'(\lambda)=2-\lambda=0
\quad \Rightarrow \quad
\lambda^*=2.
$$

The dual value is

$$
q(2)=4-2=2.
$$

The primal feasible point is $x^*=0$, and its cost is

$$
\frac{1}{2}(0-2)^2=2.
$$

The primal and dual optimal values are equal.

```python
import sympy as sp

x, lambda_multiplier = sp.symbols('x lambda')

cost = sp.Rational(1, 2) * (x - 2)**2
lagrangian = cost + lambda_multiplier * x
x_lambda = sp.solve(sp.diff(lagrangian, x), x)[0]
dual_function = sp.simplify(lagrangian.subs(x, x_lambda))
lambda_solution = sp.solve(sp.diff(dual_function, lambda_multiplier), lambda_multiplier)[0]

result = {
    'x_lambda': x_lambda,
    'dual_function': dual_function,
    'lambda_solution': lambda_solution,
    'dual_value': dual_function.subs(lambda_multiplier, lambda_solution),
    'primal_value': cost.subs(x, 0),
}

for label, value in result.items():
    print(f"{label} =", value)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

opti = ca.Opti()
decision_variable = opti.variable()
opti.minimize(0.5 * (decision_variable - 2) ** 2)
opti.subject_to(decision_variable <= 0)
opti.solver("ipopt", {"print_time": False}, {"print_level": 0, "sb": "yes"})
solution = opti.solve()

print("minimizer =", ca.DM(solution.value(decision_variable)))
print("primal_value =", ca.DM(solution.value(opti.f)))
```
