### 1.1.6.4. Injective, Surjective, and Bijective Transformations


$$
\operatorname{Ker}(T) = \{ \mathbf{0} \}
\Longleftrightarrow
T \text{ is injective}
$$

$$
\operatorname{Im}(T) = W
\Longleftrightarrow
T \text{ is surjective}
$$


**Explanation:**

Injective means different inputs stay different, surjective means every output is reached, and bijective means both properties hold at once. For a linear transformation, injectivity is read through the [kernel](./03_kernel.ipynb) and surjectivity through the [image space](./02_image_space.ipynb), so bijectivity is the point where both viewpoints agree.

**Example:**

For

$$
T(x, y, z) =
\begin{bmatrix}
x \\
z
\end{bmatrix},
$$

the transformation is surjective onto $\mathbb{R}^2$ but not injective because vectors that differ only in their $y$-component share the same output.


```python
import sympy as sp

transformation_matrix = sp.Matrix([[1, 0, 0], [0, 0, 1]])
rank_value = transformation_matrix.rank()
nullity_value = transformation_matrix.cols - rank_value

print("rank_value =", rank_value)
print("nullity_value =", nullity_value)
print("injective =", nullity_value == 0)
print("surjective =", rank_value == transformation_matrix.rows)
```
