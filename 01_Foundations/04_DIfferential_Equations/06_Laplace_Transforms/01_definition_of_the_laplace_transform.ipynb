### 1.4.6.1. Definition of the Laplace Transform

$$
\mathcal{L}\{f(t)\} = \int_0^{\infty} e^{-st} f(t)\,dt = F(s),
\qquad
\mathcal{L}\{\alpha f + \beta g\} = \alpha F(s) + \beta G(s) .
$$

**Explanation:**

The Laplace transform turns a function $f(t)$ (for $t \ge 0$) into a function $F(s)$ via an [improper integral](../../03_Calculus/03_Single_Variable_Integration/06_improper_integrals.ipynb) with the kernel $e^{-st}$; it converges for $s$ larger than some threshold. It is a **linear** operator, so it distributes over sums and constant multiples. Its power is operational: it converts differentiation in $t$ into multiplication by $s$, turning a differential equation into an algebraic one. A short table of transforms of the basic functions, built once from the definition, then serves every later computation.

**Properties:**
- Basic transforms: $\mathcal{L}\{1\} = \tfrac1s$, $\mathcal{L}\{t^n\} = \tfrac{n!}{s^{n+1}}$, $\mathcal{L}\{e^{at}\} = \tfrac{1}{s-a}$, $\mathcal{L}\{\sin kt\} = \tfrac{k}{s^2+k^2}$, $\mathcal{L}\{\cos kt\} = \tfrac{s}{s^2+k^2}$.
- Existence is guaranteed for functions that are piecewise continuous and of exponential order.

**Numerical Example:**

**From the definition.** Compute $\mathcal{L}\{1\}$ directly:

$$
\mathcal{L}\{1\} = \int_0^{\infty} e^{-st}\,dt = \left.\frac{-e^{-st}}{s}\right|_0^{\infty} = \frac{0 - (-1)}{s} = \frac{1}{s}, \qquad s > 0,
$$

since $e^{-st} \to 0$ as $t \to \infty$ when $s > 0$. Similarly $\mathcal{L}\{e^{-3t}\} = \int_0^\infty e^{-(s+3)t}\,dt = \tfrac{1}{s+3}$, and two integrations by parts give $\mathcal{L}\{\sin 2t\} = \tfrac{2}{s^2+4}$.

**Linearity.** Combining table entries,

$$
\mathcal{L}\{20e^{-3t} + 7t - 9\} = \frac{20}{s+3} + \frac{7}{s^2} - \frac{9}{s} .
$$

```python
import sympy as sp

t, s = sp.symbols("t s", positive=True)
a, k = sp.symbols("a k")

from_definition = sp.integrate(sp.exp(-s * t) * 1, (t, 0, sp.oo))
basic = {name: sp.laplace_transform(function, t, s, noconds=True)
         for name, function in {"1": sp.Integer(1), "t": t, "e^{at}": sp.exp(a * t),
                                 "sin kt": sp.sin(k * t), "cos kt": sp.cos(k * t)}.items()}
combination = sp.laplace_transform(20 * sp.exp(-3 * t) + 7 * t - 9, t, s, noconds=True)

print("L{1} from definition =", from_definition)
for name, transform in basic.items():
    print(f"L{{{name}}} = {transform}")
print("L{20 e^-3t + 7t - 9} =", combination)
```
