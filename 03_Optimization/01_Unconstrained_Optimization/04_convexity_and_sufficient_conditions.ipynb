### 3.1.4. Convexity and Sufficient Conditions


$$
\begin{aligned}
f(\theta x + (1-\theta)y)
&\le \theta f(x) + (1-\theta)f(y),\quad 0 \le \theta \le 1, \\
\nabla f(x^*) = 0\ \text{and}\ f\ \text{convex}
&\Longrightarrow x^* \text{ is a global minimum}, \\
\nabla f(x^*) = 0,\ \nabla^2 f(x^*) \succ 0
&\Longrightarrow x^* \text{ is a strict local minimum}.
\end{aligned}
$$


**Explanation:**

Convexity is the main structure that turns local information into a global conclusion. For a differentiable convex cost, the first-order condition $\nabla f(x^*)=0$ means that every direction has nonnegative first-order cost change, and the convexity inequality prevents a lower valley from appearing elsewhere. Without convexity, the same stationarity condition may occur at maxima, flat inflection points, or saddle-like points. Positive definiteness of the Hessian gives a different kind of result: it proves a strict local minimum through the curvature encoded by [positive definite matrices](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/17_positive_definite_matrix.ipynb), but it does not automatically prove global optimality.

The convexity inequality says the cost at a weighted average of two points is no larger than the same weighted average of their costs. The remaining lines say that, for convex $f$, the stationary condition $\nabla f(x^*)=0$ is enough for global optimality, and a [positive definite Hessian](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/17_positive_definite_matrix.ipynb) gives a strict local certificate using [matrix definiteness](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/18_matrix_definiteness.ipynb).

This notebook therefore separates two ideas that beginners often blend together. Convexity is a global shape assumption, while a positive definite Hessian at one point is a local curvature statement. Both ideas matter because practical nonlinear programming needs to know when a local condition is enough and when it is only a local diagnostic.

**Intuition:**

The first figure illustrates why a convex function cannot have a merely local minimum that is not global. The second figure reminds us that nonconvex level sets can surround a global minimizer in a way that defeats simple geometric intuition. The code cell then checks a quadratic whose Hessian is positive definite, making the stationary point a strict minimum.

**Numerical Example:**

The convex-level-set figure corresponds to the case where local stationarity is enough. Consider
$$
f(x_1,x_2)=2x_1^2+x_1x_2+x_2^2-4x_1.
$$

The gradient equations are
$$
\begin{aligned}
\frac{\partial f}{\partial x_1}&=4x_1+x_2-4=0,\\
\frac{\partial f}{\partial x_2}&=x_1+2x_2=0.
\end{aligned}
$$

From $x_1+2x_2=0$, $x_1=-2x_2$. Substitute into the first equation:
$$
4(-2x_2)+x_2-4=0
\quad\Longrightarrow\quad
-7x_2=4
\quad\Longrightarrow\quad
x_2=-\frac{4}{7},\qquad x_1=\frac{8}{7}.
$$

The Hessian is
$$
\nabla^2 f=\begin{bmatrix}4&1\\1&2\end{bmatrix},
\qquad 4>0,\qquad \det(\nabla^2 f)=8-1=7>0.
$$

Thus the Hessian is positive definite, the quadratic is convex, and the stationary point $(8/7,-4/7)$ is the global minimum.


```python
import sympy as sp

first_variable, second_variable = sp.symbols("x_1 x_2")
objective = 2 * first_variable**2 + first_variable * second_variable + second_variable**2 - 4 * first_variable
hessian = sp.hessian(objective, (first_variable, second_variable))
stationary_solution = sp.solve(
    [sp.diff(objective, first_variable), sp.diff(objective, second_variable)],
    (first_variable, second_variable),
)

print("hessian =", hessian)
print("stationary_solution =", stationary_solution)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

decision_variable = ca.SX.sym("x", 2)
objective = 2 * decision_variable[0] ** 2 + decision_variable[0] * decision_variable[1] + decision_variable[1] ** 2 - 4 * decision_variable[0]
gradient = ca.gradient(objective, decision_variable)
hessian = ca.jacobian(gradient, decision_variable)
hessian_value = ca.evalf(ca.substitute(hessian, decision_variable, ca.DM.zeros(2)))
stationary_point = ca.solve(hessian_value, -ca.evalf(ca.substitute(gradient, decision_variable, ca.DM.zeros(2))))

print("hessian =", hessian_value)
print("stationary_point =", stationary_point)
```
