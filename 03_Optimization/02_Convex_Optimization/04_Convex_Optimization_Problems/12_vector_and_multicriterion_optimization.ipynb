### 3.2.4.12. Vector and Multicriterion Optimization


$$\min_x^{(K)} f_0(x)$$


**Explanation:**

Vector optimization compares objective vectors through a cone order or through scalarized tradeoffs. A feasible point is efficient when no other feasible point improves all objectives in the chosen order.

The superscript $(K)$ means the objective values are compared with a cone order rather than the usual scalar order. The function $f_0(x)$ returns a vector, and minimizing it means searching for feasible points that cannot be improved in the [cone](../02_Convex_Sets/03_cones_and_conic_combinations.ipynb) order.

**Properties:**
- Cone order defines vector dominance.
- Scalarization converts selected tradeoffs into scalar convex problems.
- Efficient points need not be unique.

**Intuition:**

The figure above shows tradeoff geometry; the calculation checks one componentwise dominance relation.

<img src="../../../Figures/03020412_bv_fig_4_9_scalarization.jpeg" alt="Boyd and Vandenberghe Figure 4.9: scalarization and Pareto optimal values" style="width: 60%; height: auto;">


**Numerical Example:**

The figure above shows tradeoff geometry. Compare
$$
f(x)=\begin{bmatrix}1\\4\end{bmatrix},
\qquad
f(y)=\begin{bmatrix}3\\5\end{bmatrix}
$$
using $K=\mathbf{R}_+^2$ for minimization.

Compute
$$
f(y)-f(x)=\begin{bmatrix}3-1\\5-4\end{bmatrix}=\begin{bmatrix}2\\1\end{bmatrix}.
$$

Both components are nonnegative, so $f(x)\preceq_K f(y)$.


```python
import sympy as sp

objective_x = sp.Matrix([1, 4])
objective_y = sp.Matrix([3, 5])
difference = objective_y - objective_x

difference, [entry >= 0 for entry in difference]
```
