### 3.2.2.10. Separating and Supporting Hyperplanes


$$a^Tx\leq b\leq a^Ty$$


**Explanation:**

Separation and support turn convex geometry into certificates. A separating hyperplane distinguishes a point or set from a [convex set](01_affine_sets_and_convex_sets.ipynb), while a supporting hyperplane touches a convex set at the boundary without cutting through it.

The vector $a$ is the normal of the separating hyperplane and $b$ is its offset. The two inner products compare how far $x$ and $y$ lie along the normal direction, using the same geometry as a hyperplane.

**Properties:**
- A hyperplane satisfies one affine equality.
- A halfspace satisfies one affine inequality.
- Finite intersections of halfspaces and hyperplanes form polyhedra.

**Intuition:**

The figure above emphasizes the normal-vector side test; the arithmetic applies that test to one point.

<img src="../../../Figures/03020210_bv_fig_2_19_separating_hyperplane.jpeg" alt="Boyd and Vandenberghe Figure 2.19: separating hyperplane between two convex sets" style="width: 60%; height: auto;">


**Numerical Example:**

The figure above emphasizes the normal-vector side test; the arithmetic applies that test to one point.

Let
$$
a=\begin{bmatrix}1\\2\end{bmatrix},\quad
b=5,\quad
x=\begin{bmatrix}1\\2\end{bmatrix}.
$$

Compute the halfspace test:
$$
a^Tx=1\cdot1+2\cdot2=1+4=5.
$$

Since
$$
5\leq5,
$$
$x$ is feasible. Equality holds, so $x$ lies on the boundary hyperplane.


```python
import sympy as sp

normal_vector = sp.Matrix([1, 2])
point = sp.Matrix([1, 2])
boundary_value = sp.Integer(5)
inner_product = (normal_vector.T * point)[0]

print("inner_product =", inner_product)
print("inner_product <= boundary_value =", inner_product <= boundary_value)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

normal_vector = ca.DM([1, 2])
point = ca.DM([1, 2])
boundary_value = 5
inner_product = ca.dot(normal_vector, point)

print("inner_product =", inner_product)
print("inner_product <= boundary_value =", bool(inner_product <= boundary_value))
```
