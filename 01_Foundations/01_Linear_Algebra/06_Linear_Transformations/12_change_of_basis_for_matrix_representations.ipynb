### 1.1.6.12. Change of Basis for Matrix Representations


$$
\left[ T(\mathbf{v}) \right]_{B_W} =
{}_{B_W}\!\left[ M_T \right]_{B_V}
\left[ \mathbf{v} \right]_{B_V}
$$


**Explanation:**

Savov emphasizes that a matrix is not the transformation itself but its coefficients relative to chosen input and output bases. This notebook keeps that basis-dependence as its only concept and links back to [change of basis for vectors](../05_Geometrical_Aspects_of_Linear_Algebra/12_change_of_basis.ipynb). The special square-matrix case is separated into the next notebook on [similarity transformations](./13_similarity_transformation.ipynb).

**Example:**

Changing the basis used for both the input and output spaces replaces the old coefficient matrix with a new one that predicts the same geometric output using different coordinates.


```python
import sympy as sp

matrix_in_old_basis = sp.Matrix([[1, 0], [0, -1]])
change_of_basis_matrix = sp.Matrix(
    [
        [1 / sp.sqrt(2), 1 / sp.sqrt(2)],
        [1 / sp.sqrt(2), -1 / sp.sqrt(2)],
    ]
)
matrix_in_new_basis = sp.simplify(change_of_basis_matrix.inv() * matrix_in_old_basis * change_of_basis_matrix)

print("matrix_in_new_basis =")
sp.pprint(matrix_in_new_basis)
```
