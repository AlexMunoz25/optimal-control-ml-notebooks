### 3.3.3. Penalty and Elimination Approaches

$$
F^k(x)=f(x)+\frac{k}{2}\|h(x)\|^2
$$

$$
\nabla f(x^*)+\nabla h(x^*)\lambda^*=0
$$

**Explanation:**

The penalty approach temporarily ignores the equality constraint and adds a large cost for violating it. The first-order conditions for the penalized problems produce quantities of the form $k h(x^k)$, and their limits become Lagrange multipliers. The elimination approach removes variables using the constraint, reducing the problem to an unconstrained one; both routes explain why constrained stationarity resembles an unconstrained first-order condition after the feasible geometry is accounted for.

**Assumptions:**
- The equality constraints are smooth near the local minimum.
- The regularity condition supports eliminating variables locally, typically through an implicit-function argument.

**Numerical Example:**

Consider

$$
\min (x_1-2)^2+(x_2-1)^2
\quad \text{subject to} \quad x_1+x_2=1.
$$

Eliminate $x_2$ from the constraint:

$$
x_2=1-x_1.
$$

The reduced cost is

$$
\phi(x_1)=(x_1-2)^2+((1-x_1)-1)^2
=(x_1-2)^2+x_1^2.
$$

Differentiate and set to zero:

$$
\phi'(x_1)=2(x_1-2)+2x_1=4x_1-4=0,
$$

so

$$
x_1^*=1, \qquad x_2^*=0.
$$

Now compute the multiplier. The gradients are

$$
\nabla f(x^*)=\begin{bmatrix}2(1-2)\\2(0-1)\end{bmatrix}
=\begin{bmatrix}-2\\-2\end{bmatrix},
\qquad
\nabla h(x^*)=\begin{bmatrix}1\\1\end{bmatrix}.
$$

The stationarity equation gives

$$
\begin{bmatrix}-2\\-2\end{bmatrix}
+\lambda\begin{bmatrix}1\\1\end{bmatrix}=0,
$$

hence

$$
\lambda^*=2.
$$

The eliminated unconstrained problem and the multiplier equation identify the same constrained solution.

```python
import sympy as sp

x_1, x_2, lambda_multiplier = sp.symbols('x_1 x_2 lambda')

cost = (x_1 - 2)**2 + (x_2 - 1)**2
constraint = x_1 + x_2 - 1
reduced_cost = cost.subs(x_2, 1 - x_1)
stationary_coordinate = sp.solve(sp.diff(reduced_cost, x_1), x_1)[0]
candidate = {x_1: stationary_coordinate, x_2: 1 - stationary_coordinate}

gradient_cost = sp.Matrix([sp.diff(cost, x_1), sp.diff(cost, x_2)]).subs(candidate)
gradient_constraint = sp.Matrix([sp.diff(constraint, x_1), sp.diff(constraint, x_2)]).subs(candidate)
lambda_solution = sp.solve(gradient_cost + lambda_multiplier * gradient_constraint, [lambda_multiplier], dict=True)

result = {
    'reduced_cost': sp.expand(reduced_cost),
    'candidate': candidate,
    'gradient_cost': gradient_cost,
    'lambda_solution': lambda_solution,
}
result
```
