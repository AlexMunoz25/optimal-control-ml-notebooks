### 1.1.7.18. Matrix Definiteness


$$
\begin{aligned}
\mathbf{v}^{\mathsf{T}} A \mathbf{v} > 0 \quad \forall\, \mathbf{v} \neq \mathbf{0} &\;\Longrightarrow\; \text{positive definite (PD)} \\
\mathbf{v}^{\mathsf{T}} A \mathbf{v} \geq 0 \quad \forall\, \mathbf{v} &\;\Longrightarrow\; \text{positive semidefinite (PSD)} \\
\mathbf{v}^{\mathsf{T}} A \mathbf{v} < 0 \quad \forall\, \mathbf{v} \neq \mathbf{0} &\;\Longrightarrow\; \text{negative definite (ND)} \\
\mathbf{v}^{\mathsf{T}} A \mathbf{v} \leq 0 \quad \forall\, \mathbf{v} &\;\Longrightarrow\; \text{negative semidefinite (NSD)}
\end{aligned}
$$


**Explanation:**

The definiteness of a matrix classifies the sign of the quadratic form $\mathbf{v}^{\mathsf{T}} A \mathbf{v}$ over all nonzero vectors $\mathbf{v}$. The strictly positive case—positive definite—is treated in [positive definite matrices](./10_positive_definite_matrix.ipynb); this notebook covers the full spectrum. Definiteness is completely determined by the signs of the eigenvalues: all positive gives PD, all nonnegative gives PSD, all negative gives ND, all nonpositive gives NSD. When eigenvalues have mixed signs the matrix is indefinite, which corresponds geometrically to a saddle-shaped quadratic form.

**Properties:**

- PD: all eigenvalues $\lambda_i > 0$.
- PSD: all eigenvalues $\lambda_i \geq 0$; rank deficiency is possible.
- ND: all eigenvalues $\lambda_i < 0$.
- NSD: all eigenvalues $\lambda_i \leq 0$.
- Indefinite: eigenvalues of mixed sign.
- $A$ is PSD if and only if $A = B^{\mathsf{T}} B$ for some matrix $B$.

**Example:**

The three matrices

$$
A_{\text{PSD}} =
\begin{bmatrix} 1 & 0 \\ 0 & 0 \end{bmatrix},
\quad
A_{\text{ND}} =
\begin{bmatrix} -2 & 0 \\ 0 & -3 \end{bmatrix},
\quad
A_{\text{indef}} =
\begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix}
$$

have eigenvalues $\{1, 0\}$, $\{-2, -3\}$, and $\{1, -1\}$, confirming PSD, ND, and indefinite classification respectively.


```python
import sympy as sp

psd_matrix = sp.Matrix([[1, 0], [0, 0]])
negative_definite_matrix = sp.Matrix([[-2, 0], [0, -3]])
indefinite_matrix = sp.Matrix([[1, 0], [0, -1]])

print("PSD eigenvalues        =", sorted(psd_matrix.eigenvals()))
print("ND eigenvalues         =", sorted(negative_definite_matrix.eigenvals()))
print("indefinite eigenvalues =", sorted(indefinite_matrix.eigenvals()))
```

```python
import numpy as np
import matplotlib.pyplot as plt

grid_points = np.linspace(-2, 2, 200)
v1_grid, v2_grid = np.meshgrid(grid_points, grid_points)

psd_form = v1_grid ** 2
negative_definite_form = -2 * v1_grid ** 2 - 3 * v2_grid ** 2
indefinite_form = v1_grid ** 2 - v2_grid ** 2

sample_vector = np.array([1.0, 2.0])
sample_psd_value = sample_vector[0] ** 2
sample_negative_definite_value = -2 * sample_vector[0] ** 2 - 3 * sample_vector[1] ** 2
sample_indefinite_value = sample_vector[0] ** 2 - sample_vector[1] ** 2

print("sample_vector =", sample_vector)
print("PSD quadratic value =", sample_psd_value)
print("negative definite quadratic value =", sample_negative_definite_value)
print("indefinite quadratic value =", sample_indefinite_value)

fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].contourf(v1_grid, v2_grid, psd_form, levels=20, cmap="coolwarm")
axes[0].set_title("PSD: $v_1^2$")
axes[0].set_xlabel("$v_1$")
axes[0].set_ylabel("$v_2$")

axes[1].contourf(v1_grid, v2_grid, negative_definite_form, levels=20, cmap="coolwarm")
axes[1].set_title("ND: $-2v_1^2 - 3v_2^2$")
axes[1].set_xlabel("$v_1$")
axes[1].set_ylabel("$v_2$")

axes[2].contourf(v1_grid, v2_grid, indefinite_form, levels=20, cmap="coolwarm")
axes[2].set_title("Indefinite: $v_1^2 - v_2^2$")
axes[2].set_xlabel("$v_1$")
axes[2].set_ylabel("$v_2$")

plt.tight_layout()
plt.show()
```
