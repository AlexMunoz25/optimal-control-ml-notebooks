### 1.1.9.8. Kronecker Delta

$$
\delta^i{}_j =
\begin{cases} 1 & i = j \\ 0 & i \neq j \end{cases}
$$

$$
\delta^i{}_j A^j = A^i, \qquad \delta^i{}_j \delta^j{}_k = \delta^i{}_k, \qquad \delta^i{}_i = n
$$

**Explanation:**

- The Kronecker delta is the identity mixed tensor.
- Contracting with $\delta^i{}_j$ leaves the other tensor unchanged.
- The self-contraction $\delta^i{}_i$ counts the dimension of the space.
- In Cartesian coordinates it coincides with the metric tensor.

**Connections:**

- This is the tensor version of the identity matrix from the matrix modules.
- It appears whenever direct and inverse Jacobians cancel during a contraction.

**Example:**

- For $A^j = (2, 5, 7)$, $\delta^i{}_j A^j = A^i$.
- In three dimensions, $\delta^i{}_i = 3$.


```python
import numpy as np

kronecker_delta = np.eye(3)
vector_components = np.array([2.0, 5.0, 7.0])

identity_action = np.einsum('ij,j->i', kronecker_delta, vector_components)
self_contraction = np.einsum('ii->', kronecker_delta)

print("kronecker_delta =\n", kronecker_delta)
print("identity_action =", identity_action)
print("self_contraction =", self_contraction)
```
