### 1.1.8.4. Active Transformation


$$
\mathbf{A}' = R_{\alpha} \mathbf{A}
$$


**Explanation:**

An active transformation rotates the vector itself while the coordinate axes stay fixed. Comparing it with the [passive transformation](./03_passive_transformation.ipynb) shows why similar-looking matrices can describe different physical actions depending on what is held fixed.

<p align="center">
  <img src="../../../Figures/01010801_active_transformation_matrix.jpeg" alt="Active transformation matrix" width="420">
</p>

**Example:**

Rotating the vector $(5, 3)$ by $150^{\circ}$ in a fixed basis produces a new geometric vector with new coordinates in that same basis.


```python
import numpy as np

rotation_angle = np.deg2rad(150.0)
active_matrix = np.array(
    [
        [np.cos(rotation_angle), -np.sin(rotation_angle)],
        [np.sin(rotation_angle), np.cos(rotation_angle)],
    ]
)
vector_components = np.array([5.0, 3.0])
active_components = active_matrix @ vector_components

print("active_components =", np.round(active_components, 4))
```
