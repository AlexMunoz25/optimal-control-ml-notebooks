### 1.3.5.6. Linearization and the Tangent Plane

$$
z = f(x_0, y_0) + f_x(x_0, y_0)(x - x_0) + f_y(x_0, y_0)(y - y_0),
$$

$$
f(\mathbf{x}) \approx f(\mathbf{x}_0) + \nabla f(\mathbf{x}_0)^{\top}(\mathbf{x} - \mathbf{x}_0) .
$$

**Explanation:**

The tangent plane is the multivariable [linear approximation](../02_Single_Variable_Differentiation/10_linear_approximation.ipynb): near a base point $\mathbf{x}_0$, the surface is replaced by the affine function whose value and [gradient](./03_gradient.ipynb) match $f$ there. In vector form it is the first-order Taylor expansion, accurate for small displacements and degrading as $\|\mathbf{x} - \mathbf{x}_0\|$ grows. This is exactly the operation that linearizes nonlinear dynamics about an operating point in control and produces the local linear models used in [Gauss–Newton](../../../03_Optimization/01_Unconstrained_Optimization/17_gauss_newton_method.ipynb) and trust-region methods.

**Properties:**
- The plane matches $f$ to first order: same value and same partials at $\mathbf{x}_0$.
- The normal vector to the surface $z = f(x,y)$ at the point is $\langle f_x, f_y, -1\rangle$.

**Intuition:**

Near the base point the tangent plane approximates the surface, the two-variable analogue of the tangent line.

**Numerical Example:**

Linearize $f(x,y) = x^2 + y^2$ at $(x_0, y_0) = (1, 1)$. There $f(1,1) = 2$ and $\nabla f(1,1) = [2, 2]^\top$, so

$$
z = 2 + 2(x - 1) + 2(y - 1) = 2x + 2y - 2 .
$$

Estimating $f(1.1, 0.9)$: the plane gives $2(1.1) + 2(0.9) - 2 = 2.0$, while the true value is $1.1^2 + 0.9^2 = 1.21 + 0.81 = 2.02$ — an error of $0.02$, small because the displacement from $(1,1)$ is small.

```python
import sympy as sp

x, y = sp.symbols("x y")
f = x**2 + y**2
base_point = {x: 1, y: 1}

gradient = sp.Matrix([sp.diff(f, x), sp.diff(f, y)])
tangent_plane = (f.subs(base_point)
                 + gradient[0].subs(base_point) * (x - 1)
                 + gradient[1].subs(base_point) * (y - 1))

estimate = tangent_plane.subs({x: sp.Rational(11, 10), y: sp.Rational(9, 10)})
true_value = f.subs({x: sp.Rational(11, 10), y: sp.Rational(9, 10)})

print("tangent plane z =", sp.expand(tangent_plane))
print("plane estimate at (1.1, 0.9) =", estimate, "=", float(estimate))
print("true f(1.1, 0.9)             =", true_value, "=", float(true_value))
print("error                        =", float(true_value - estimate))
```

**Equivalent CasADi implementation:**

CasADi builds the same first-order model from the value and gradient evaluated at the base point.

```python
import casadi as ca

variables = ca.SX.sym("x", 2)
f = variables[0]**2 + variables[1]**2
model = ca.Function("model", [variables], [f, ca.gradient(f, variables)])

base_point = ca.DM([1, 1])
base_value, base_gradient = model(base_point)
query_point = ca.DM([1.1, 0.9])
plane_estimate = base_value + ca.dot(base_gradient, query_point - base_point)

print("plane estimate at (1.1, 0.9) =", float(plane_estimate))
print("true f(1.1, 0.9)             =", float(model(query_point)[0]))
```
