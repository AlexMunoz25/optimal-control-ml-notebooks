### 3.2.2.11. Dual Cones and Dual Generalized Inequalities


$$K^*=\{y\mid y^Tx\geq0\ \text{for all }x\in K\}$$


**Explanation:**

The dual cone contains the linear functionals that are nonnegative on the original cone. It converts cone membership into scalar inequalities, which is why dual [cones](03_cones_and_conic_combinations.ipynb) appear naturally in [generalized inequalities](09_generalized_inequalities_and_proper_cones.ipynb) and conic duality.

The dual cone $K^*$ contains all vectors $y$ that make a nonnegative [inner product](../../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/02_inner_product_space.ipynb) with every $x$ in $K$. This definition turns membership in the original cone into scalar inequalities.

**Properties:**
- The cone defines the meaning of inequality.
- For the nonnegative orthant, every component must be nonnegative.
- This order supports vector optimization and conic duality.

**Numerical Example:**

The figure above represents cone order geometrically; the arithmetic checks the difference vector.

Let
$$
x=\begin{bmatrix}1\\2\end{bmatrix},\quad
y=\begin{bmatrix}3\\5\end{bmatrix},\quad K=\mathbf{R}_+^2.
$$

The order test is $x\preceq_K y$ if $y-x\in K$. Compute
$$
y-x=\begin{bmatrix}3-1\\5-2\end{bmatrix}=\begin{bmatrix}2\\3\end{bmatrix}.
$$

Both components are nonnegative, so $x\preceq_K y$.


```python
import sympy as sp

point_x = sp.Matrix([1, 2])
point_y = sp.Matrix([3, 5])
difference = point_y - point_x
nonnegative = [entry >= 0 for entry in difference]

print("difference =", difference)
print("nonnegative =", nonnegative)
```
