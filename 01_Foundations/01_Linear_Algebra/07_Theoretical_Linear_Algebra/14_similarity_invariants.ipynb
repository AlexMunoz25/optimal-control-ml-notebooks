### 1.1.7.14. Similarity Invariants


$$
\det(A) = \prod_i \lambda_i,
\qquad
\operatorname{tr}(A) = \sum_i \lambda_i
$$


**Explanation:**

Determinant and trace do not change under similarity transformations. Their invariance follows from the change-of-basis viewpoint of [similarity transformations](../06_Linear_Transformations/13_similarity_transformation.ipynb), and the eigenvalue interpretation explains why these quantities are basis independent.

**Example:**

For

$$
A =
\begin{bmatrix}
9 & -2 \\
-2 & 6
\end{bmatrix},
$$

the determinant is $50$ and the trace is $15$, matching the product and sum of the eigenvalues $5$ and $10$.


```python
import sympy as sp

matrix_a = sp.Matrix([[9, -2], [-2, 6]])
eigenvalues = list(matrix_a.eigenvals().keys())

print("determinant_value =", matrix_a.det())
print("trace_value =", matrix_a.trace())
print("product_of_eigenvalues =", sp.prod(eigenvalues))
print("sum_of_eigenvalues =", sum(eigenvalues))
```
