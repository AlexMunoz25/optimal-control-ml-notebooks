### 1.3.6.6. Laplacian

$$
\Delta f = \nabla^{2} f = \operatorname{div}(\nabla f) = \nabla\cdot\nabla f
= \frac{\partial^{2} f}{\partial x^{2}} + \frac{\partial^{2} f}{\partial y^{2}} + \frac{\partial^{2} f}{\partial z^{2}} .
$$

**Explanation:**

The Laplacian is the [divergence](./05_divergence.ipynb) of the [gradient](../05_Multivariable_Differential_Calculus/03_gradient.ipynb) — the sum of unmixed second partials, equivalently the trace of the [Hessian](../05_Multivariable_Differential_Calculus/09_hessian_matrix.ipynb). It measures how much a function's value at a point exceeds the average of its neighbours: $\Delta f < 0$ at a peak, $\Delta f > 0$ at a dip, and $\Delta f = 0$ (a *harmonic* function) where the value equals the local average. The Laplacian is the spatial operator of the heat/diffusion equation and the Fokker–Planck equation, and appears as a smoothness regularizer in machine learning.

**Properties:**
- $\Delta f = \operatorname{tr}(H_f)$, the trace of the Hessian.
- Harmonic functions ($\Delta f = 0$) satisfy the mean-value property and have no interior extrema.

**Numerical Example:**

For the paraboloid $f(x,y) = x^2 + y^2$: $f_{xx} = 2$, $f_{yy} = 2$, so

$$
\Delta f = 2 + 2 = 4 > 0 ,
$$

positive everywhere — each point sits below its neighbourhood average, consistent with the upward bowl. For $g(x,y) = x^2 - y^2$: $g_{xx} = 2$, $g_{yy} = -2$, so

$$
\Delta g = 2 + (-2) = 0 ,
$$

$g$ is harmonic — its saddle shape balances curvature exactly.

```python
import sympy as sp

x, y = sp.symbols("x y")

def laplacian(scalar_field, variables):
    return sum(sp.diff(scalar_field, variable, 2) for variable in variables)

paraboloid = x**2 + y**2
saddle = x**2 - y**2

print("Δ(x^2 + y^2) =", laplacian(paraboloid, (x, y)), " (> 0: bowl)")
print("Δ(x^2 - y^2) =", laplacian(saddle, (x, y)), " (= 0: harmonic)")
```

**Equivalent CasADi implementation:**

CasADi computes the Laplacian as the trace of the Hessian of the scalar field.

```python
import casadi as ca

variables = ca.SX.sym("x", 2)
paraboloid = variables[0]**2 + variables[1]**2
hessian, _ = ca.hessian(paraboloid, variables)

print("Δ(x^2 + y^2) = tr(H) =", float(ca.evalf(ca.trace(hessian))))
```
