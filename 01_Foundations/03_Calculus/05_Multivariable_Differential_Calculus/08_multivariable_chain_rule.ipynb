### 1.3.5.8. Multivariable Chain Rule

$$
\frac{d}{dt} f\big(x(t), y(t)\big) = f_x\,\frac{dx}{dt} + f_y\,\frac{dy}{dt}
= \nabla f \cdot \frac{d\mathbf{x}}{dt};
\qquad
\frac{\partial}{\partial \mathbf{x}}\,\mathbf{g}\big(\mathbf{h}(\mathbf{x})\big) = J_{\mathbf{g}}\,J_{\mathbf{h}} .
$$

**Explanation:**

The multivariable chain rule differentiates a composition by summing the contributions through every intermediate variable — for a scalar field along a curve it is the [inner product](../../01_Linear_Algebra/02_Vector/04_vector_dot_product.ipynb) of the [gradient](./03_gradient.ipynb) with the velocity. In full generality the derivative of a composition of vector maps is the **product of their [Jacobians](./07_jacobian.ipynb)**, evaluated along the chain. Applied layer by layer this matrix-product form is exactly [backpropagation](../07_Matrix_Calculus_and_Automatic_Differentiation/08_reverse_mode_automatic_differentiation.ipynb) and the adjoint method used to differentiate through dynamics.

**Properties:**
- Generalized form: if $z = f(x_1,\ldots,x_n)$ and each $x_i = x_i(t)$, then $\tfrac{dz}{dt} = \sum_i \tfrac{\partial f}{\partial x_i}\tfrac{dx_i}{dt}$.
- The Jacobian-product rule $J_{g\circ h} = J_g\,J_h$ reduces to $f'(g(x))g'(x)$ in one dimension.

**Numerical Example:**

Let $f(x, y) = x^2 y$ with $x(t) = \cos t$ and $y(t) = \sin t$. The partials are $f_x = 2xy$, $f_y = x^2$, and the velocities are $x'(t) = -\sin t$, $y'(t) = \cos t$:

$$
\frac{df}{dt} = (2xy)(-\sin t) + (x^2)(\cos t)
= 2\cos t\sin t(-\sin t) + \cos^2 t\,\cos t = -2\cos t\sin^2 t + \cos^3 t .
$$

At $t = 0$ ($x = 1$, $y = 0$): $\tfrac{df}{dt} = -2(1)(0) + (1) = 1$, which matches differentiating the substituted form $f = \cos^2 t\sin t$ directly.

```python
import sympy as sp

t = sp.symbols("t")
x, y = sp.symbols("x y")
f = x**2 * y
x_of_t, y_of_t = sp.cos(t), sp.sin(t)

chain_rule = (sp.diff(f, x).subs({x: x_of_t, y: y_of_t}) * sp.diff(x_of_t, t)
              + sp.diff(f, y).subs({x: x_of_t, y: y_of_t}) * sp.diff(y_of_t, t))
direct = sp.diff(f.subs({x: x_of_t, y: y_of_t}), t)
difference = sp.simplify(chain_rule - direct)

print("chain rule df/dt        =", sp.simplify(chain_rule))
print("direct substitution     =", sp.simplify(direct))
print("difference (should be 0) =", difference)
print("df/dt at t=0            =", chain_rule.subs(t, 0))
```

**Equivalent CasADi implementation:**

Composing the curve into $f$ as one CasADi graph and differentiating reproduces $df/dt$ via the Jacobian-product chain.

```python
import casadi as ca

t = ca.SX.sym("t")
x_of_t, y_of_t = ca.cos(t), ca.sin(t)
composed = x_of_t**2 * y_of_t
chain_derivative = ca.Function("dfdt", [t], [ca.gradient(composed, t)])

print("df/dt at t=0 =", float(chain_derivative(0.0)))
```
