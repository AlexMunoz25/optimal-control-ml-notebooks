### 3.2.11.9. Primal-Dual Search Direction


$$r_{\rm cent}=-\operatorname{diag}(\lambda)f(x)-(1/t)\mathbf1$$


**Explanation:**

Primal-dual interior-point methods solve perturbed KKT equations directly. Their search direction is a Newton step for primal feasibility, dual feasibility, and centering residuals. Feasibility and optimality improve together as the residuals shrink.

The vector $f(x)$ stacks inequality residuals, and $\lambda$ stacks their multipliers. The centering residual compares componentwise products $\lambda_i f_i(x)$ with the target value $-1/t$.

**Properties:**
- The centering residual measures perturbed complementarity error.
- The parameter t sets the target complementarity level.
- The search direction is computed from the linearized residual equations.

**Intuition:**

The example computes one scalar centering residual.


**Numerical Example:**

For one inequality, suppose
$$
f(x)=-1,\qquad \lambda=2,\qquad t=4.
$$

The centering residual is
$$
r_{\rm cent}=-\lambda f(x)-\frac1t.
$$

Substitute the values:
$$
r_{\rm cent}=-2(-1)-\frac14=2-\frac14=\frac74.
$$

A nonzero residual means the current point is not exactly centered.


```python
import sympy as sp

constraint_value = sp.Integer(-1)
multiplier = sp.Integer(2)
t_parameter = sp.Integer(4)
centering_residual = -multiplier * constraint_value - 1 / t_parameter

centering_residual
```
