### A5.1.6. Build Dependency Injection Container from Scratch

**Problem:**

Build a container that registers factory functions for dependencies and resolves them by name, automatically injecting sub-dependencies.

**Explanation:**

A dependency injection container is a registry of factory functions. When you ask for a dependency by name, the container calls the registered factory. The factory itself can ask the container for other dependencies, forming a chain of automatic resolution.

How to build it:

1. Store a dictionary mapping names to factory callables in `__init__`.
2. `register(name, factory)`: store the factory under the given name. The factory takes the container as its only argument.
3. `resolve(name)`: look up the factory by name and call it, passing `self` (the container) so the factory can resolve its own dependencies.

**Example:**

Register a `"database"` factory and a `"user_service"` factory. The user service factory calls `container.resolve("database")` to get its dependency. Calling `container.resolve("user_service")` automatically creates the database first.

```python
class Container:
    def __init__(self):
        self.factories = {}

    def register(self, name, factory):
        self.factories[name] = factory

    def resolve(self, name):
        return self.factories[name](self)


class Database:
    def __init__(self, connection_string):
        self.connection_string = connection_string


class UserService:
    def __init__(self, database):
        self.database = database


container = Container()
container.register("database", lambda c: Database("sqlite://app.db"))
container.register("user_service", lambda c: UserService(c.resolve("database")))

service = container.resolve("user_service")
print(f"UserService created with database: {service.database.connection_string}")
```
