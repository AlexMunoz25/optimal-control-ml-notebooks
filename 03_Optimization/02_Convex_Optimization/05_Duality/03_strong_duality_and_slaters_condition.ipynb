### 3.2.5.3. Strong Duality and Slater Condition


$$d^\star=p^\star$$


**Explanation:**

Strong duality means the best dual lower bound reaches the primal optimum. Slater condition is a common convex constraint qualification that guarantees this equality when a strictly feasible point exists.

**Properties:**
- Weak duality gives a lower bound for minimization.
- A primal feasible point gives an upper bound.
- A zero gap proves optimality when both sides are feasible.

**Intuition:**

The figure above gives the certificate geometry; the arithmetic computes the gap between the certificates.

<img src="../../../Figures/03020503_bv_fig_5_6_slater_strong_duality.jpeg" alt="Boyd and Vandenberghe Figure 5.6: strong duality proof under Slater condition" style="width: 60%; height: auto;">


**Numerical Example:**

The figure above gives the certificate geometry; the arithmetic computes the gap between the certificates.

Suppose a primal feasible point has value $p=4$ and a dual feasible point has value $d=3$. Weak duality gives
$$
d\leq p^\star\leq p.
$$

Substitute the values:
$$
3\leq p^\star\leq4.
$$

The certificate gap is
$$
p-d=4-3=1.
$$

The dual point proves that no feasible primal solution can have value below $3$, while the primal point proves the optimum is at most $4$.


```python
import sympy as sp

primal_value = sp.Integer(4)
bound_value = sp.Integer(3)
gap = primal_value - bound_value

bound_value <= primal_value, gap
```
