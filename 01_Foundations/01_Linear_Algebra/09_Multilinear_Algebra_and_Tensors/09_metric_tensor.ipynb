### 1.1.9.9. Metric Tensor

$$
g_{ij} = \vec{e}_i \cdot \vec{e}_j, \qquad ds^2 = g_{ij}\,dx^i\,dx^j
$$

$$
|\vec{A}|^2 = g_{ij} A^i A^j = g^{ij} A_i A_j = A^i A_i
$$

$$
\cos\theta = \frac{g_{ij} A^i B^j}{\sqrt{g_{ij} A^i A^j}\,\sqrt{g_{ij} B^i B^j}}
$$

**Explanation:**

- The metric tensor stores the dot products of the coordinate basis vectors.
- It turns coordinate increments into invariant distances and tensor components into invariant lengths and angles.
- In Cartesian coordinates the metric is the identity; in curvilinear coordinates it can vary from point to point.

**Connections:**

- This extends the earlier notebooks on generalized inner products, non-orthogonal coordinates, and the Jacobian matrix.
- In ML, a positive-definite matrix inside a quadratic form or Mahalanobis distance plays the same algebraic role as a fixed metric.

**Example:**

- In spherical coordinates, the Jacobian columns are the basis vectors associated with $r$, $\theta$, and $\phi$.
- Their pairwise dot products produce
$$
g_{ij} =
\begin{bmatrix}
1 & 0 & 0 \\
0 & r^2 & 0 \\
0 & 0 & r^2 \sin^2 \theta
\end{bmatrix}.
$$
- The diagonal entries give the scale factors $h_1 = 1$, $h_2 = r$, and $h_3 = r \sin \theta$.


```python
import sympy as sp

radial_coordinate, polar_angle, azimuth_angle = sp.symbols('r theta phi', positive=True)
cartesian_coordinates = [
    radial_coordinate * sp.sin(polar_angle) * sp.cos(azimuth_angle),
    radial_coordinate * sp.sin(polar_angle) * sp.sin(azimuth_angle),
    radial_coordinate * sp.cos(polar_angle),
]
spherical_coordinates = [radial_coordinate, polar_angle, azimuth_angle]
jacobian_rows = [
    [
        sp.diff(cartesian_coordinate, spherical_coordinate)
        for spherical_coordinate in spherical_coordinates
    ]
    for cartesian_coordinate in cartesian_coordinates
]
coordinate_jacobian = sp.Matrix(jacobian_rows)
metric_tensor_covariant = sp.simplify(coordinate_jacobian.T * coordinate_jacobian)
metric_tensor_contravariant = sp.simplify(metric_tensor_covariant.inv())
scale_factors = [
    sp.sqrt(metric_tensor_covariant[diagonal_index, diagonal_index])
    for diagonal_index in range(3)
]

print("metric_tensor_covariant =")
sp.pprint(metric_tensor_covariant)
print("metric_tensor_contravariant =")
sp.pprint(metric_tensor_contravariant)
print("scale_factors =", [sp.simplify(scale_factor) for scale_factor in scale_factors])
```
