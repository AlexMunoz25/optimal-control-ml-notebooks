### 1.4.1.2. Initial-Value Problems

$$
\frac{dy}{dx} = f(x, y), \qquad y(x_0) = y_0 .
$$

**Explanation:**

A function $\phi$ is a **solution** of an ODE on an interval $I$ if it has the required derivatives and reduces the equation to an identity there. Solving an ODE usually produces a **general solution**: a family of solutions carrying arbitrary constants — exactly the constant of integration $C$ that appears when an [antiderivative](../../03_Calculus/03_Single_Variable_Integration/01_antiderivative_and_indefinite_integral.ipynb) is taken, one constant per order. An **initial-value problem (IVP)** pairs the ODE with **initial conditions** that fix those constants, selecting one **particular solution**. Geometrically, the general solution is a family of curves and the initial condition $y(x_0) = y_0$ picks out the single curve passing through $(x_0, y_0)$.

**Intuition:**

The general solution is a whole family of curves; each initial condition selects one member of the family.

**Numerical Example:**

Consider $y' = y$. Separating and integrating (the method is developed in the next section) gives the general solution

$$
\frac{dy}{y} = dx \;\Rightarrow\; \ln|y| = x + C_1 \;\Rightarrow\; y = C e^{x},
$$

a one-parameter family. Now impose two different initial conditions.

**(a)** $y(0) = 3$: substitute $x = 0$, $y = 3$ into $y = Ce^x$:
$$
3 = C e^{0} = C \;\Rightarrow\; C = 3 \;\Rightarrow\; y = 3e^{x}.
$$

**(b)** $y(1) = -2$: substitute $x = 1$, $y = -2$:
$$
-2 = C e^{1} \;\Rightarrow\; C = -2e^{-1} \;\Rightarrow\; y = -2e^{-1}e^{x} = -2e^{\,x-1}.
$$

Each initial condition turns the family $Ce^x$ into one particular solution curve.

```python
import sympy as sp

x = sp.symbols("x")
y = sp.Function("y")
ode = sp.Eq(y(x).diff(x), y(x))

general_solution = sp.dsolve(ode, y(x))
particular_a = sp.dsolve(ode, y(x), ics={y(0): 3})
particular_b = sp.dsolve(ode, y(x), ics={y(1): -2})

print("general solution :", general_solution)
print("with y(0) = 3    :", particular_a)
print("with y(1) = -2   :", particular_b)
print("check y(0)=3     :", particular_a.rhs.subs(x, 0))
print("check y(1)=-2    :", sp.simplify(particular_b.rhs.subs(x, 1)))
```
