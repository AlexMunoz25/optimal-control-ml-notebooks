### 1.1.8.8. Covariant Components


$$
\mathbf{A} = A_i \mathbf{e}^{\,i},
\qquad
A_i = \mathbf{A} \cdot \mathbf{e}_i,
\qquad
A_{i'} = \frac{\partial x^j}{\partial x^{i'}} A_j
$$


**Explanation:**

Covariant components are associated with the perpendicular-projection viewpoint, but in a skew basis they do not combine with the original basis vectors. Instead, the lower-index components pair with the dual basis so that the vector is reconstructed as $\mathbf{A} = A_i \mathbf{e}^{\,i}$. These components transform with the same law as the original basis vectors, which is why they are called covariant.

</p>

**Example:**

For
$$
\mathbf{A} =
\begin{bmatrix}
7 \\
2
\end{bmatrix},
\qquad
\mathbf{e}_1 =
\begin{bmatrix}
1 \\
3
\end{bmatrix},
\qquad
\mathbf{e}_2 =
\begin{bmatrix}
4 \\
0
\end{bmatrix},
$$
the covariant components are
$$
A_1 = \mathbf{A} \cdot \mathbf{e}_1 = 13,
\qquad
A_2 = \mathbf{A} \cdot \mathbf{e}_2 = 28.
$$


```python
import sympy as sp

basis_vector_1 = sp.Matrix([1, 3])
basis_vector_2 = sp.Matrix([4, 0])
vector_a = sp.Matrix([7, 2])
primal_basis_matrix = sp.Matrix.hstack(basis_vector_1, basis_vector_2)
dual_basis_matrix = primal_basis_matrix.inv().T
covariant_components = sp.Matrix([vector_a.dot(basis_vector_1), vector_a.dot(basis_vector_2)])
reconstructed_vector = (
    covariant_components[0] * dual_basis_matrix[:, 0]
    + covariant_components[1] * dual_basis_matrix[:, 1]
)

print("covariant_components =", covariant_components)
print("reconstructed_vector =", sp.simplify(reconstructed_vector))
```
