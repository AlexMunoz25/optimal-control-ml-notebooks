### 1.1.5.17. Subspace Bases via RREF


$$
A \xrightarrow{\operatorname{rref}} R
$$


**Explanation:**

Basis extraction for subspaces can be reduced to row reduction. Pivot columns recover a basis for the column space, nonzero rows of the reduced matrix give a basis for the row space, and free-variable directions describe the null space. This turns the structural picture from the [fundamental spaces](./16_matrix_fundamental_spaces.ipynb) into a computation based on [reduced row echelon form](../04_Computational_Linear_Algebra/03_reduced_row_echelon_form.ipynb).

**Example:**

For

$$
A=\begin{bmatrix}4&-4&0\\1&1&-2\\2&-6&4\end{bmatrix},
$$

row reduction gives

$$
\operatorname{rref}(A)=\begin{bmatrix}1&0&-1\\0&1&-1\\0&0&0\end{bmatrix}.
$$

The nonzero rows form a row-space basis:

$$
\left\{\begin{bmatrix}1&0&-1\end{bmatrix},\begin{bmatrix}0&1&-1\end{bmatrix}\right\}.
$$

The pivot columns are columns $1$ and $2$, so the corresponding original columns form a column-space basis:

$$
\left\{\begin{bmatrix}4\\1\\2\end{bmatrix},\begin{bmatrix}-4\\1\\-6\end{bmatrix}\right\}.
$$


```python
import sympy as sp

matrix_a = sp.Matrix([[4, -4, 0], [1, 1, -2], [2, -6, 4]])
reduced_matrix, pivot_columns = matrix_a.rref()
row_space_basis = [reduced_matrix.row(row_index) for row_index in range(reduced_matrix.rows) if any(reduced_matrix.row(row_index))]
column_space_basis = [matrix_a.col(column_index) for column_index in pivot_columns]

print("matrix_a =")
sp.pprint(matrix_a)
print("reduced_matrix =")
sp.pprint(reduced_matrix)
print("pivot_columns =", pivot_columns)
print("row_space_basis =", row_space_basis)
print("column_space_basis =", column_space_basis)
```
