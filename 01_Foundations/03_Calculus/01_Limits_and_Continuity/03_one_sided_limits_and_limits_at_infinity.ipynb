### 1.3.1.3. One-Sided Limits and Limits at Infinity

$$
\lim_{x \to a^{-}} f(x), \qquad \lim_{x \to a^{+}} f(x), \qquad
\lim_{x \to \infty} f(x) = L .
$$

$$
\lim_{x \to a} f(x) = L \iff \lim_{x \to a^{-}} f(x) = \lim_{x \to a^{+}} f(x) = L .
$$

**Explanation:**

A one-sided limit restricts the approach to a single side: $x \to a^{-}$ uses values below $a$, and $x \to a^{+}$ uses values above $a$. The two-sided limit exists exactly when both one-sided limits exist and agree, which is why a jump between the two sides makes the limit fail to exist. A limit at infinity, $\lim_{x\to\infty} f(x) = L$, describes the function's end behaviour — the horizontal asymptote it settles toward — and is the language used for the growth and convergence rates studied in [asymptotic notation](./05_asymptotic_notation_big_o_little_o.ipynb).

**Properties:**
- For a rational function, $\lim_{x\to\pm\infty}$ is the ratio of leading coefficients when numerator and denominator have equal degree, $0$ when the denominator degree is larger, and $\pm\infty$ when the numerator degree is larger.
- An infinite limit such as $\lim_{x\to 0^{+}} 1/x = +\infty$ signals a vertical asymptote, not a (finite) limit value.

**Numerical Example:**

Let $f(x) = \dfrac{|x|}{x}$. For $x > 0$, $f(x) = 1$; for $x < 0$, $f(x) = -1$. Hence

$$
\lim_{x \to 0^{+}} \frac{|x|}{x} = 1, \qquad \lim_{x \to 0^{-}} \frac{|x|}{x} = -1 .
$$

The one-sided limits disagree, so $\lim_{x\to 0} f(x)$ **does not exist** (a jump discontinuity).

For end behaviour, take $g(x) = \dfrac{2x^2}{x^2 + 1}$. Dividing numerator and denominator by $x^2$,

$$
\lim_{x \to \infty} \frac{2x^2}{x^2 + 1} = \lim_{x \to \infty} \frac{2}{1 + 1/x^2} = \frac{2}{1 + 0} = 2,
$$

so $y = 2$ is a horizontal asymptote.

```python
import sympy as sp

x = sp.symbols("x", real=True)
sign_function = sp.Abs(x) / x
rational_function = 2 * x**2 / (x**2 + 1)

right_limit = sp.limit(sign_function, x, 0, "+")
left_limit = sp.limit(sign_function, x, 0, "-")
two_sided_exists = right_limit == left_limit
end_behaviour = sp.limit(rational_function, x, sp.oo)

print("lim x->0+  |x|/x =", right_limit)
print("lim x->0-  |x|/x =", left_limit)
print("two-sided limit exists:", two_sided_exists)
print("lim x->oo  2x^2/(x^2+1) =", end_behaviour)
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, (left_axis, right_axis) = plt.subplots(1, 2, figsize=(11, 4))

negative_points = np.linspace(-2, -0.02, 100)
positive_points = np.linspace(0.02, 2, 100)
left_axis.plot(negative_points, np.sign(negative_points), color="C0", linewidth=2)
left_axis.plot(positive_points, np.sign(positive_points), color="C0", linewidth=2)
left_axis.scatter([0, 0], [-1, 1], facecolors="none", edgecolors="red", s=80, zorder=5)
left_axis.set_title("|x|/x: jump at 0 (limit DNE)")
left_axis.set_xlabel("x"); left_axis.set_ylabel("f(x)"); left_axis.grid(True, alpha=0.3)

far_points = np.linspace(0.1, 40, 400)
right_axis.plot(far_points, 2 * far_points**2 / (far_points**2 + 1), color="C1", linewidth=2)
right_axis.axhline(2, color="gray", linestyle="--", label="asymptote y=2")
right_axis.set_title(r"$2x^2/(x^2+1) \to 2$ as $x\to\infty$")
right_axis.set_xlabel("x"); right_axis.set_ylabel("g(x)"); right_axis.legend(); right_axis.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
