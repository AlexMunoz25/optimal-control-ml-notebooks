### 1.1.9.2. Tensor Rank and Shape

$$
\operatorname{rank}(T) = n,
\qquad
\operatorname{shape}(T) = (d_1, d_2, \ldots, d_n),
\qquad
\#(T) = d_1 d_2 \cdots d_n
$$

**Explanation:**

Rank counts the number of index positions needed to identify one component of a tensor, while shape records how many values each index can take in a chosen array representation. In Fleisch's three-dimensional setting, a rank-2 tensor has $3^2 = 9$ components and a rank-3 tensor has $3^3 = 27$. This is the same bookkeeping used in [Vector Definition](../02_Vector/01_vector_definition.ipynb) and [Matrix Definition](../03_Matrix/01_matrix_definition.ipynb), extended to more axes.

**Example:**

If
$$
T_{ijk},
\qquad
i, j, k \in \{1, 2, 3\},
$$
then
$$
\operatorname{rank}(T) = 3,
\qquad
\operatorname{shape}(T) = (3, 3, 3),
\qquad
\#(T) = 27.
$$
A collection of $m$ such tensors would be stored with array shape $(m, 3, 3, 3)$, but each individual sample is still rank 3.

```python
import numpy as np

vector_tensor = np.zeros(3)
matrix_tensor = np.zeros((3, 3))
rank_3_tensor = np.zeros((3, 3, 3))
sampled_matrix_tensor = np.zeros((5, 3, 3))

tensor_examples = [
    ("vector_tensor", vector_tensor),
    ("matrix_tensor", matrix_tensor),
    ("rank_3_tensor", rank_3_tensor),
    ("sampled_matrix_tensor", sampled_matrix_tensor),
]

for tensor_name, tensor_value in tensor_examples:
    print(
        f"{tensor_name:<21} rank = {tensor_value.ndim}, "
        f"shape = {tensor_value.shape}, components = {tensor_value.size}"
    )
```
