### 3.2.7.5. Nonparametric Distribution Estimation


$$p_i\geq0,
\qquad\sum_i p_i=1$$


**Explanation:**

Nonparametric distribution estimation treats the probabilities themselves as optimization variables. The probability simplex is convex, and additional moment or likelihood constraints can be added without leaving convex optimization when they are affine or convex. Feasibility is nonnegativity plus normalization.

The variables $p_i$ are probabilities assigned to discrete outcomes. Nonnegativity and the sum-to-one equation make the vector $p$ a probability distribution over those outcomes.

**Properties:**
- Every probability must be nonnegative.
- The probabilities must sum to one.
- Affine expectation constraints preserve convexity.

**Intuition:**

The example checks simplex feasibility directly from the probability vector entries.


**Numerical Example:**

Let
$$
p=\begin{bmatrix}1/4\\1/2\\1/4\end{bmatrix}.
$$

Check nonnegativity:
$$
1/4\geq0,\qquad 1/2\geq0,\qquad 1/4\geq0.
$$

Check normalization:
$$
1/4+1/2+1/4=1/4+2/4+1/4=1.
$$

Therefore $p$ is feasible in the probability simplex.


```python
import sympy as sp

probabilities = sp.Matrix([sp.Rational(1, 4), sp.Rational(1, 2), sp.Rational(1, 4)])
normalization = sum(probabilities)
nonnegative = [probability >= 0 for probability in probabilities]

print("normalization =", normalization)
print("nonnegative =", nonnegative)
```
