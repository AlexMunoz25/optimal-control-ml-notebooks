### 1.1.7.12. Projection Matrix as an Idempotent Matrix


$$
\Pi^2 = \Pi
$$


**Explanation:**

A projection matrix is an idempotent matrix: applying it twice gives the same result as applying it once. The geometric definition already appeared in [projection matrices](../05_Geometrical_Aspects_of_Linear_Algebra/09_projection_matrix.ipynb), so the emphasis here is the special-matrix viewpoint and the consequence that the eigenvalues are $0$ or $1$.

**Example:**

For the projection onto the $x$-axis,

$$
\Pi =
\begin{bmatrix}
1 & 0 \\
0 & 0
\end{bmatrix},
$$

and $\Pi^2 = \Pi$.


```python
import sympy as sp

projection_matrix = sp.Matrix([[1, 0], [0, 0]])

print("square_matrix =")
sp.pprint(projection_matrix * projection_matrix)
print("is_idempotent =", projection_matrix * projection_matrix == projection_matrix)
```
