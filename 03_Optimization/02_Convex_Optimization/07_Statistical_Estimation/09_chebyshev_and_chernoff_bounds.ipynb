### 3.2.7.9. Chebyshev and Chernoff Bounds


$$\operatorname{prob}(|X-\mu|\geq a)\leq\sigma^2/a^2$$


**Explanation:**

Chebyshev and Chernoff bounds turn partial distribution information into probability guarantees. Convex optimization can search over distributions, moments, or exponential parameters to make such bounds tight under stated information. Feasibility encodes known moment or support constraints.

**Properties:**
- Chebyshev bounds use mean and variance.
- Chernoff bounds use exponential moments.
- The result is a conservative probability guarantee.

**Intuition:**

The example evaluates the Chebyshev bound using only variance and a deviation threshold.


**Numerical Example:**

Let $\mu=0$, $\sigma=2$, and $a=5$. Chebyshev's inequality gives
$$
\operatorname{prob}(|X-\mu|\geq a)\leq\frac{\sigma^2}{a^2}.
$$

Substitute the values:
$$
\frac{\sigma^2}{a^2}=\frac{2^2}{5^2}=\frac4{25}.
$$

Thus
$$
\operatorname{prob}(|X|\geq5)\leq\frac4{25}.
$$


```python
import sympy as sp

standard_deviation = sp.Integer(2)
deviation_threshold = sp.Integer(5)
bound = standard_deviation ** 2 / deviation_threshold ** 2

bound
```
