### 1.1.7.3. Generalized Inner Product


$$
\langle A, B \rangle_{\mathrm{HS}} = \operatorname{tr}(A^{\mathsf{T}} B)
$$


**Explanation:**

A generalized inner product shows that the inner-product axioms apply beyond ordinary coordinate vectors. The Hilbert-Schmidt product is a concrete example, and the abstract axioms themselves are introduced in [inner product spaces](./14_inner_product_space.ipynb).

**Example:**

For two matrices, the Hilbert-Schmidt inner product multiplies matching entries and sums the result through the trace formula.


```python
import sympy as sp

matrix_a = sp.Matrix([[1, 0], [0, 0]])
matrix_b = sp.Matrix([[0, 0], [0, 1]])
hilbert_schmidt_value = (matrix_a.T * matrix_b).trace()

print("hilbert_schmidt_value =", hilbert_schmidt_value)
```
