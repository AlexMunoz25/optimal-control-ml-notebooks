### 1.1.3.14. Matrix Outer Product

$$
\mathbf{u} \otimes \mathbf{v} = \mathbf{u}\mathbf{v}^T
$$

**Explanation:**

The outer product of vectors $\mathbf{u} \in \mathbb{R}^m$ and $\mathbf{v} \in \mathbb{R}^n$ produces an $m \times n$ matrix.
The resulting matrix has rank 1 (if both vectors are non-zero), and every row is a scaled copy of $\mathbf{v}^T$.
Useful for constructing matrices from features and weights in ML.

Builds on [Vectors](../02_Vector/01_vector_definition.ipynb) and contrasts with the scalar-valued [Dot Product](../02_Vector/04_vector_dot_product.ipynb). The result is a matrix in the sense of [Matrices](./01_matrix_definition.ipynb).

**Example:**

Let $\mathbf{u} = [1, 2]^T$ and $\mathbf{v} = [3, 4, 5]^T$. Then

$$
\mathbf{u} \otimes \mathbf{v} =
\begin{bmatrix}
1 \\
2
\end{bmatrix}
\begin{bmatrix}
3 & 4 & 5
\end{bmatrix}
=
\begin{bmatrix}
3 & 4 & 5 \\
6 & 8 & 10
\end{bmatrix}.
$$


```python
import sympy as sp

vector_u = sp.Matrix([1, 2])
vector_v = sp.Matrix([3, 4, 5])

outer_product = vector_u * vector_v.T

print("u ⊗ v =")
sp.pprint(outer_product)
```
