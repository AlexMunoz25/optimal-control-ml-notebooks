### 3.9.30. Finite Difference Derivative Approximations

$$
\frac{\partial f(x)}{\partial x_i}
\approx \frac{f(x+he_i)-f(x)}{h},
\qquad
\frac{\partial f(x)}{\partial x_i}
\approx \frac{f(x+he_i)-f(x-he_i)}{2h}.
$$


**Explanation:**

Nonderivative methods begin with the practical fact that derivatives may be unavailable, unreliable, or too expensive to derive analytically. The calculus and automatic-differentiation foundation notebook on [numerical, symbolic, and automatic differentiation](../../01_Foundations/03_Calculus/07_Matrix_Calculus_and_Automatic_Differentiation/09_numerical_symbolic_and_automatic_differentiation.ipynb) owns the general derivative-computation comparison. In optimization, finite differences matter because they let us approximate or check gradients of an objective using only cost evaluations.

Forward differences are cheaper but have first-order truncation error, while central differences require more function evaluations but have second-order truncation error in exact arithmetic. It is also important that $h$ cannot simply be made arbitrarily small, because roundoff error grows when nearly equal objective values are subtracted.

The vector $e_i$ is the coordinate basis direction for the $i$th variable. The forward formula compares $f$ at $x$ and $x+he_i$, while the centered formula compares symmetric perturbations $x\pm he_i$ to approximate the same partial derivative.

**Intuition:**

Finite differences are especially useful for simulation-defined objectives: one may know how to evaluate a cost but not have an exact formula for its gradient. The code cell compares forward and central approximations of a scalar objective gradient at the current iterate, then converts that approximation into the local descent direction.


**Numerical Example:**

Suppose a one-dimensional optimization routine can evaluate
$$
f(x)=\frac12(x-3)^2+\sin x
$$
but wants a derivative estimate at the current iterate $x=1$. The exact derivative is
$$
f'(x)=x-3+\cos x,
\qquad
f'(1)=-2+\cos 1\approx -1.459698.
$$

With $h=0.001$, the forward estimate is
$$
\frac{f(1.001)-f(1)}{0.001},
$$
and the central estimate is
$$
\frac{f(1.001)-f(0.999)}{2(0.001)}.
$$

Both approximate the objective gradient. The corresponding local descent direction is the negative derivative, so a negative gradient value tells the method to increase $x$ from the current point.


```python
import sympy as sp

variable = sp.symbols("x")
objective = sp.Rational(1, 2) * (variable - 3) ** 2 + sp.sin(variable)
exact_derivative_expression = sp.diff(objective, variable)

point = sp.Integer(1)
stepsize = sp.Rational(1, 1000)
forward_difference = ((objective.subs(variable, point + stepsize) - objective.subs(variable, point)) / stepsize).evalf()
central_difference = ((objective.subs(variable, point + stepsize) - objective.subs(variable, point - stepsize)) / (2 * stepsize)).evalf()
exact_derivative = exact_derivative_expression.subs(variable, point)
descent_direction = -central_difference

print("objective =", objective)
print("exact_derivative_expression =", exact_derivative_expression)
print("forward_difference =", forward_difference)
print("central_difference =", central_difference)
print("exact_derivative =", exact_derivative, "=", exact_derivative.evalf())
print("descent_direction_from_central_difference =", descent_direction)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

variable = ca.SX.sym("x")
objective = 0.5 * (variable - 3) ** 2 + ca.sin(variable)
objective_function = ca.Function("f", [variable], [objective])
exact_derivative_function = ca.Function("df", [variable], [ca.gradient(objective, variable)])

point = 1.0
stepsize = 1e-3
forward_difference = float(objective_function(point + stepsize) - objective_function(point)) / stepsize
central_difference = float(objective_function(point + stepsize) - objective_function(point - stepsize)) / (2 * stepsize)
exact_derivative = exact_derivative_function(point)
descent_direction = -central_difference

print("forward_difference =", forward_difference)
print("central_difference =", central_difference)
print("exact_derivative =", exact_derivative)
print("descent_direction_from_central_difference =", descent_direction)
```
