### 3.3.14. Least-Squares and Norm Approximation


$$\min_x \|Ax-b\|$$

$$\|r\|_2^2,\qquad\|r\|_1,\qquad\|r\|_\infty$$

**Explanation:**

The norm approximation problem chooses decision variables so the residual $r=Ax-b$ is small in a specified [norm](../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb). The selected norm keeps the objective convex and fixes the geometry of the error measure, so a single problem $\min_x\|Ax-b\|$ becomes least-squares, $\ell_1$, or Chebyshev approximation depending on the norm. Squared two-norm emphasizes larger errors, one-norm sums absolute errors, and infinity norm controls the worst residual.

The vector $r$ is the residual. The three expressions measure it with squared Euclidean norm, one-norm, and infinity norm, so the same residual can penalize total squared error, total absolute error, or worst absolute error.

**Properties:**
- The same residual vector can have different objective values.
- The infinity norm is the maximum absolute residual.
- Norm and penalty choices encode robustness or smoothness preferences.

**Intuition:**

The example evaluates three common residual measures on one vector.

**Numerical Example:**

Let
$$
r=\begin{bmatrix}1\\-2\\3\end{bmatrix}.
$$

The squared two-norm is
$$
\|r\|_2^2=1^2+(-2)^2+3^2=1+4+9=14.
$$

The one-norm is
$$
\|r\|_1=|1|+|-2|+|3|=1+2+3=6.
$$

The infinity norm is
$$
\|r\|_\infty=\max\{1,2,3\}=3.
$$


```python
import sympy as sp

residual = sp.Matrix([1, -2, 3])
squared_l2 = sum(entry ** 2 for entry in residual)
l1_norm = sum(abs(entry) for entry in residual)
linfinity_norm = max(abs(entry) for entry in residual)

print("squared_l2 =", squared_l2)
print("l1_norm =", l1_norm)
print("linfinity_norm =", linfinity_norm)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

residual = ca.DM([1, -2, 3])
squared_l2 = ca.dot(residual, residual)
l1_norm = ca.norm_1(residual)
linfinity_norm = ca.norm_inf(residual)

print("squared_l2 =", squared_l2)
print("l1_norm =", l1_norm)
print("linfinity_norm =", linfinity_norm)
```
