### 1.1.5.6. Distance from a Plane to the Origin


$$
d(P, \mathbf{0}) = \frac{\left| \mathbf{n} \cdot \mathbf{p}_0 \right|}{\lVert \mathbf{n} \rVert}
$$


**Explanation:**

The shortest path from the origin to a plane runs along the plane normal, so the distance is the magnitude of the scalar projection of $\mathbf{p}_0$ onto the unit normal direction. Once the geometry of a [plane in $\mathbb{R}^3$](./02_plane_in_r3.ipynb) is fixed, the computation can also be read as the normal-direction version of [projection onto a plane](./08_projection_onto_a_plane.ipynb).

**Example:**

If

$$
\mathbf{n} =
\begin{bmatrix}
-3 \\
0 \\
-4
\end{bmatrix},
\qquad
\mathbf{p}_0 =
\begin{bmatrix}
1 \\
2 \\
3
\end{bmatrix},
$$

then

$$
d(P, \mathbf{0}) =
\frac{\left| -3 - 12 \right|}{5}
= 3.
$$


```python
import sympy as sp

plane_normal = sp.Matrix([-3, 0, -4])
point_in_plane = sp.Matrix([1, 2, 3])
distance_value = sp.simplify(abs(plane_normal.dot(point_in_plane)) / plane_normal.norm())

print("plane_normal =", list(plane_normal))
print("point_in_plane =", list(point_in_plane))
print("distance_value =", distance_value)
```
