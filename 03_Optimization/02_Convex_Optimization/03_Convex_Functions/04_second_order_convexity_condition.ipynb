### 3.2.3.4. Second-Order Convexity Condition


$$\nabla^2f(x)\succeq0$$


**Explanation:**

For twice differentiable functions, convexity can be checked through curvature. A positive semidefinite Hessian means every second directional derivative is nonnegative, so the function bends upward along every line in the domain.

The display says the Hessian matrix $\nabla^2 f(x)$ is positive semidefinite at each point in the domain. In [matrix definiteness](../../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/18_matrix_definiteness.ipynb) terms, every second directional derivative is nonnegative.

**Properties:**
- All [eigenvalues](../../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/08_eigenvalue.ipynb) must be nonnegative for positive semidefiniteness.
- [Positive semidefinite matrices](../../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/18_matrix_definiteness.ipynb) form a convex [cone](../02_Convex_Sets/03_cones_and_conic_combinations.ipynb).
- Matrix inequalities use this cone as an order relation.

**Intuition:**

The example checks a concrete two-by-two matrix through its eigenvalues.


**Numerical Example:**

Let
$$
X=\begin{bmatrix}2&1\\1&2\end{bmatrix}.
$$

The characteristic polynomial is
$$
\det(X-\lambda I)=(2-\lambda)^2-1.
$$

Expanding gives
$$
(2-\lambda)^2-1=\lambda^2-4\lambda+3=(\lambda-1)(\lambda-3).
$$

Thus the eigenvalues are $1$ and $3$. Since both are nonnegative, $X\succeq0$.


```python
import sympy as sp

matrix = sp.Matrix([[2, 1], [1, 2]])
eigenvalues = matrix.eigenvals()

print("matrix =", matrix)
print("eigenvalues =", eigenvalues)
```
