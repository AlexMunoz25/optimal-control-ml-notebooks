### 1.1.7.28. Singular Value Decomposition


$$
A = U \Sigma V^{\mathsf{T}}
$$


**Explanation:**

The singular value decomposition factors a matrix into left singular directions, right singular directions, and stretching magnitudes. Unlike [eigendecomposition](./04_eigendecomposition.ipynb), it applies to rectangular matrices and is built from the spectrum of $A^{\mathsf{T}} A$.

**Example:**

For any real matrix, the diagonal entries of $\Sigma$ are the singular values, which are the square roots of the eigenvalues of $A^{\mathsf{T}} A$.


```python
import sympy as sp

matrix_a = sp.Matrix([[2, 1], [1, 2]])
matrix_u, singular_value_matrix, matrix_v = matrix_a.singular_value_decomposition()
matrix_v_transpose = matrix_v.T

print("matrix_u =")
sp.pprint(matrix_u)
print("ОЈ =")
sp.pprint(singular_value_matrix)
print("matrix_v_transpose =")
sp.pprint(matrix_v_transpose)
print("U ОЈ VбөҖ =")
sp.pprint(matrix_u * singular_value_matrix * matrix_v_transpose)
```
