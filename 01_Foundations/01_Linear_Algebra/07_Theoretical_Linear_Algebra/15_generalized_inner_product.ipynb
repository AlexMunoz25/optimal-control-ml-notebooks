### 1.1.7.15. Generalized Inner Product


$$
\langle A, B \rangle_{\mathrm{HS}} = \operatorname{tr}(A^{\mathsf{T}} B)
$$


**Explanation:**

Savov's generalized inner products show that the inner-product axioms apply beyond ordinary vectors. This notebook keeps the single idea that new spaces can receive new inner products, and uses the Hilbert-Schmidt product as the concrete example. The abstract axioms themselves live in the previous notebook on [inner product spaces](./14_inner_product_space.ipynb).

**Example:**

For two matrices, the Hilbert-Schmidt inner product multiplies matching entries and sums the result through the trace formula.


```python
import numpy as np

matrix_a = np.array([[1.0, 0.0], [0.0, 0.0]])
matrix_b = np.array([[0.0, 0.0], [0.0, 1.0]])
hilbert_schmidt_value = float(np.trace(matrix_a.T @ matrix_b))

print("hilbert_schmidt_value =", hilbert_schmidt_value)
```
