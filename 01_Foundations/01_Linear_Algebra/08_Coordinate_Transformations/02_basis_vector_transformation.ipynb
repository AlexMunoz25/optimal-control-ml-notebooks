### 1.1.8.2. Basis-Vector Transformation


$$
\begin{pmatrix}
\text{New basis} \\
\text{vectors}
\end{pmatrix}
=
\begin{pmatrix}
\text{Direct} \\
\text{transformation} \\
\text{matrix}
\end{pmatrix}
\begin{pmatrix}
\text{Original basis} \\
\text{vectors}
\end{pmatrix}
$$

$$
\begin{bmatrix}
\hat{\mathbf{i}}'_x \\
\hat{\mathbf{i}}'_y
\end{bmatrix}
=
\begin{bmatrix}
\cos \alpha & -\sin \alpha \\
\sin \alpha & \cos \alpha
\end{bmatrix}
\begin{bmatrix}
1 \\
0
\end{bmatrix},
\qquad
\begin{bmatrix}
\hat{\mathbf{j}}'_x \\
\hat{\mathbf{j}}'_y
\end{bmatrix}
=
\begin{bmatrix}
\cos \alpha & -\sin \alpha \\
\sin \alpha & \cos \alpha
\end{bmatrix}
\begin{bmatrix}
0 \\
1
\end{bmatrix}
$$


**Explanation:**

A basis-vector transformation rotates the basis directions themselves and then expresses those new basis vectors in the original coordinate system. The direct transformation matrix therefore acts on the old basis vectors to produce the new basis. For an orthogonal rotation, the columns of that matrix are exactly the rotated basis vectors written in the original frame.

  <img src="../../../Figures/01010801_rotated_basis_in_original_system.jpeg" alt="Rotated basis vectors expressed in the original system" width="680">
</p>

**Example:**

For a $150^{\circ}$ counterclockwise rotation,
$$
\hat{\mathbf{i}}' =
\begin{bmatrix}
\cos 150^{\circ} \\
\sin 150^{\circ}
\end{bmatrix}
=
\begin{bmatrix}
-0.866 \\
0.500
\end{bmatrix},
\qquad
\hat{\mathbf{j}}' =
\begin{bmatrix}
-\sin 150^{\circ} \\
\cos 150^{\circ}
\end{bmatrix}
=
\begin{bmatrix}
-0.500 \\
-0.866
\end{bmatrix}.
$$


```python
import numpy as np

rotation_angle_degrees = 150.0
rotation_angle = np.deg2rad(rotation_angle_degrees)
direct_transformation_matrix = np.array(
    [
        [np.cos(rotation_angle), -np.sin(rotation_angle)],
        [np.sin(rotation_angle), np.cos(rotation_angle)],
    ]
)
original_basis_i = np.array([1.0, 0.0])
original_basis_j = np.array([0.0, 1.0])
rotated_basis_i = direct_transformation_matrix @ original_basis_i
rotated_basis_j = direct_transformation_matrix @ original_basis_j

print("direct_transformation_matrix =")
print(np.round(direct_transformation_matrix, 4))
print("rotated_basis_i =", np.round(rotated_basis_i, 4))
print("rotated_basis_j =", np.round(rotated_basis_j, 4))
```
