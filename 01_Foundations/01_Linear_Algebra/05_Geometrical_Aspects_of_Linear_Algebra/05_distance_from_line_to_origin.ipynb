### 1.1.5.5. Distance from a Line to the Origin


$$
d(\ell, \mathbf{0}) =
\left\lVert
\mathbf{p}_0 -
\frac{\mathbf{p}_0 \cdot \mathbf{v}}{\lVert \mathbf{v} \rVert^2} \mathbf{v}
\right\rVert
$$


**Explanation:**

The distance from a line to the origin is found by removing from the anchor point $\mathbf{p}_0$ the component parallel to the line direction $\mathbf{v}$. The remaining perpendicular component has exactly the length of the shortest segment from the origin to the line. The formula therefore relies on both the geometry of a [line in $\mathbb{R}^3$](./01_line_in_r3.ipynb) and the idea of [projection onto a line](./07_projection_onto_a_line.ipynb).

**Example:**

Let a line pass through

$$
\mathbf{p}=\begin{bmatrix}4\\5\\6\end{bmatrix}
$$

with direction

$$
\mathbf{v}=\begin{bmatrix}1\\0\\1\end{bmatrix}.
$$

The component of $\mathbf{p}$ parallel to the line is

$$
\frac{\mathbf{p}\cdot\mathbf{v}}{\mathbf{v}\cdot\mathbf{v}}\mathbf{v}
=
\frac{4\cdot1+5\cdot0+6\cdot1}{1^2+0^2+1^2}
\begin{bmatrix}1\\0\\1\end{bmatrix}
=
5\begin{bmatrix}1\\0\\1\end{bmatrix}
=
\begin{bmatrix}5\\0\\5\end{bmatrix}.
$$

The perpendicular component is

$$
\begin{bmatrix}4\\5\\6\end{bmatrix}-\begin{bmatrix}5\\0\\5\end{bmatrix}=\begin{bmatrix}-1\\5\\1\end{bmatrix}.
$$

Its length is

$$
\sqrt{(-1)^2+5^2+1^2}=\sqrt{27}=3\sqrt{3}.
$$


```python
import sympy as sp

line_point = sp.Matrix([4, 5, 6])
line_direction = sp.Matrix([1, 0, 1])
parallel_component = line_point.dot(line_direction) / line_direction.dot(line_direction) * line_direction
perpendicular_component = line_point - parallel_component
distance_value = sp.simplify(perpendicular_component.norm())

print("parallel_component =", parallel_component)
print("perpendicular_component =", perpendicular_component)
print("distance_value =", distance_value)
```
