### 3.1.11. Rate of Convergence Concepts


$$
\begin{aligned}
e(x^k) &\le q\beta^k &&\text{linear convergence}, \\
\limsup_{k\to\infty}\frac{e(x^{k+1})}{e(x^k)} &= 0 &&\text{superlinear convergence}, \\
e(x^{k+1}) &= O(e(x^k)^2) &&\text{quadratic convergence}.
\end{aligned}
$$


**Explanation:**

Convergence alone does not say whether an algorithm is useful. A method may approach a stationary point but require so many iterations that it is impractical. Bertsekas therefore introduces rate of convergence by comparing an error sequence with standard sequences, especially geometric sequences. The error can be the distance to the limit point, the cost difference, or another nonnegative measure that vanishes at the solution.

Linear convergence means that the tail of the error behaves no worse than a fixed geometric contraction. Superlinear convergence means the contraction factor itself tends to zero, and quadratic convergence means the next error is proportional to the square of the current error. These distinctions prepare the reader for the central contrast of Chapter 1: steepest descent is simple but often linearly slow, while Newton-like methods can be much faster near a nonsingular minimizer.

**Example:**

If two methods both converge, the one with the faster asymptotic rate can become dramatically more accurate after only a few additional iterations. The code cell compares representative linear, superlinear, and quadratic error sequences, not as a theorem about a specific method, but as a numerical picture of the definitions.


```python
linear_errors = [0.8**iteration for iteration in range(8)]
superlinear_errors = [0.8**(2**iteration) for iteration in range(8)]
quadratic_errors = [0.5]

for iteration in range(7):
    quadratic_errors.append(quadratic_errors[-1] ** 2)

linear_errors[-1], superlinear_errors[-1], quadratic_errors[-1]
```
