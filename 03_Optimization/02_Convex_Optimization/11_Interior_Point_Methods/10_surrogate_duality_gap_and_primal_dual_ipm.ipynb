### 3.2.11.10. Surrogate Duality Gap and Primal-Dual IPM


$$\hat\eta=-f(x)^T\lambda$$


**Explanation:**

The surrogate duality gap is a computable complementarity measure for primal-dual iterates. It uses current inequality values and multipliers even before exact complementarity holds. Small residuals together with a small surrogate gap indicate near optimality.

**Properties:**
- The gap is nonnegative for negative inequality values and nonnegative multipliers.
- The barrier parameter is often chosen from the current gap.
- The gap measures complementarity error.

**Intuition:**

The figure above shows primal-dual progress; the calculation computes the current gap.

<img src="../../../Figures/03021110_bv_fig_11_21_primal_dual_progress.jpeg" alt="Boyd and Vandenberghe Figure 11.21: primal-dual interior-point progress for an LP" style="width: 60%; height: auto;">


**Numerical Example:**

The figure above shows primal-dual progress. Let
$$
f(x)=\begin{bmatrix}-1\\-2\end{bmatrix},\qquad
\lambda=\begin{bmatrix}2\\1\end{bmatrix}.
$$

Compute the inner product:
$$
f(x)^T\lambda=(-1)\cdot2+(-2)\cdot1=-4.
$$

The surrogate duality gap is
$$
\hat\eta=-f(x)^T\lambda=-(-4)=4.
$$


```python
import sympy as sp

constraint_values = sp.Matrix([-1, -2])
multipliers = sp.Matrix([2, 1])
surrogate_gap = -(constraint_values.T * multipliers)[0]

surrogate_gap
```
