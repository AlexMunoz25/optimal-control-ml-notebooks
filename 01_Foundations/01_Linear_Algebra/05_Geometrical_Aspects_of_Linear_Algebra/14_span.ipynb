### 1.1.5.14. Span


$$
\operatorname{span}\left\{ \mathbf{v}_1, \ldots, \mathbf{v}_n \right\}
=
\left\{
\alpha_1 \mathbf{v}_1 + \cdots + \alpha_n \mathbf{v}_n
\mid
\alpha_1, \ldots, \alpha_n \in \mathbb{R}
\right\}
$$


**Explanation:**

The span of a set of vectors is the set of all linear combinations of those vectors. Geometrically, span answers the question of what space those directions generate together, and it connects directly to the idea of a [basis](./10_basis.ipynb).

**Example:**

Let

$$
\mathbf{v}_1=\begin{bmatrix}1\\2\end{bmatrix},
\quad
\mathbf{v}_2=\begin{bmatrix}1\\3\end{bmatrix},
\quad
\mathbf{w}=\begin{bmatrix}5\\11\end{bmatrix}.
$$

Solve $c_1\mathbf{v}_1+c_2\mathbf{v}_2=\mathbf{w}$:

$$
\begin{aligned}
c_1+c_2&=5,\\
2c_1+3c_2&=11.
\end{aligned}
$$

From $c_1=5-c_2$, the second equation becomes

$$
2(5-c_2)+3c_2=11,
$$

so $10+c_2=11$ and $c_2=1$. Then $c_1=4$. Therefore

$$
\mathbf{w}=4\mathbf{v}_1+\mathbf{v}_2.
$$


```python
import sympy as sp

vector_v1 = sp.Matrix([1, 2])
vector_v2 = sp.Matrix([1, 3])
target_vector = sp.Matrix([5, 11])
coefficient_vector = sp.Matrix.hstack(vector_v1, vector_v2).LUsolve(target_vector)

print("coefficient_vector =", coefficient_vector)
print("target_vector =", sp.Matrix.hstack(vector_v1, vector_v2) * coefficient_vector)
```
