### 1.1.9.6. Tensor Contraction

$$
D^j{}_l = C^{ij}{}_{il},
\qquad
\operatorname{rank}(D) = \operatorname{rank}(C) - 2,
\qquad
\frac{\partial x^i}{\partial x^k} = \delta^i{}_k
$$

**Explanation:**

A contraction sets one upper index equal to one lower index and sums over that repeated index. The upper-lower pairing is essential: it is what allows the direct and inverse transformation factors to collapse to the identity, just as in [Einstein Summation Convention](../08_Coordinate_Transformations/09_einstein_summation_convention.ipynb). [Matrix Trace](../03_Matrix/17_matrix_trace.ipynb) is the simplest example of tensor contraction.

**Example:**

If
$$
C^{ij}{}_{kl} = A^i{}_k B^j{}_l,
$$
then contracting the first upper index with the first lower index gives
$$
D^j{}_l = C^{ij}{}_{il} = A^i{}_i B^j{}_l = (\operatorname{tr} A) B^j{}_l.
$$

```python
import numpy as np

matrix_a = np.array([[1.0, 2.0], [3.0, 4.0]])
matrix_b = np.array([[5.0, 6.0], [7.0, 8.0]])
rank_4_tensor = np.einsum('ik,jl->ijkl', matrix_a, matrix_b)

matrix_a_trace = np.einsum('ii->', matrix_a)
contracted_rank_2_tensor = np.einsum('ijil->jl', rank_4_tensor)
expected_rank_2_tensor = matrix_a_trace * matrix_b

print("matrix_a_trace =", matrix_a_trace)
print("contracted_rank_2_tensor =\n", contracted_rank_2_tensor)
print("expected_rank_2_tensor =\n", expected_rank_2_tensor)
```
