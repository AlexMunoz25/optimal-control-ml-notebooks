### 3.2.4.2. Equivalent Forms and Epigraph Transformations


$$\min_x f_0(x)\Longleftrightarrow\min_{x,t}t\quad\text{subject to}\quad f_0(x)\leq t$$


**Explanation:**

Epigraph transformations introduce an auxiliary variable that stands above an objective or constraint function. Minimizing the original function can then be written as minimizing this height variable subject to an epigraph constraint.

**Properties:**
- Convex functions have convex epigraphs.
- Quasiconvex functions have convex sublevel sets.
- Epigraph transformations replace objectives by constraints.

**Intuition:**

The figure above shows set membership derived from a function value; the arithmetic checks that membership.

<img src="../../../Figures/03020402_bv_fig_4_1_epigraph_form.jpeg" alt="Boyd and Vandenberghe Figure 4.1: geometric interpretation of epigraph form" style="width: 60%; height: auto;">


**Numerical Example:**

The figure above shows set membership derived from a function value; the arithmetic checks that membership.

Let $f(x)=x^2$ and $x=2$. First compute
$$
f(2)=2^2=4.
$$

At level $5$,
$$
4\leq5,
$$
so $x=2$ is in the sublevel set. At level $3$,
$$
4\leq3
$$
is false, so $x=2$ is not in that lower sublevel set.


```python
import sympy as sp

function_value = sp.Integer(2) ** 2

function_value <= 5, function_value <= 3
```
