### 1.1.9.4. Tensor Addition

$$
C_{ij} = A_{ij} + B_{ij},
\qquad
C^{ij} = A^{ij} + B^{ij},
\qquad
C^i{}_j = A^i{}_j + B^i{}_j
$$

**Explanation:**

Two tensors can be added only when they have the same rank and the same arrangement of upper and lower indices. Then addition is carried out component by component, exactly as in [Vector Addition](../02_Vector/02_vector_addition.ipynb) and [Matrix Addition](../03_Matrix/02_matrix_addition.ipynb). Because the tensor transformation law is linear, the sum has the same structure as each addend and is therefore again a tensor.

**Example:**

Let

$$
A=\begin{bmatrix}1&2\\0&3\end{bmatrix},
\qquad
B=\begin{bmatrix}4&-1\\2&5\end{bmatrix}.
$$

Add corresponding components:

$$
A+B
=\begin{bmatrix}
1+4&2+(-1)\\
0+2&3+5
\end{bmatrix}
=\begin{bmatrix}5&1\\2&8\end{bmatrix}.
$$


```python
import sympy as sp

mixed_tensor_a = sp.Array([[1, 2], [0, 3]])
mixed_tensor_b = sp.Array([[4, -1], [2, 5]])
mixed_tensor_sum = mixed_tensor_a + mixed_tensor_b

print("mixed_tensor_sum =")
sp.pprint(mixed_tensor_sum)
```
