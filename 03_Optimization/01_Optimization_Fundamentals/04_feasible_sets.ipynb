### 3.1.4. Feasible Sets

$$
\mathcal{X} = \{\mathbf{x}\in\mathbb{R}^n \;:\; g_i(\mathbf{x}) \le 0,\; i=1,\dots,m,\quad h_j(\mathbf{x}) = 0,\; j=1,\dots,p\}.
$$

**Explanation:**

The feasible set $\mathcal{X}$ is the collection of all decision vectors that satisfy every constraint simultaneously; optimization is restricted to this set. Inequality constraints $g_i(\mathbf{x})\le 0$ carve out regions, equality constraints $h_j(\mathbf{x})=0$ restrict to surfaces, and their intersection is $\mathcal{X}$. A problem is *feasible* when $\mathcal{X}\neq\varnothing$ and *infeasible* otherwise; when $\mathcal{X}$ is convex the problem inherits the strong structure studied throughout [convex optimization](../03_Convex_Optimization/01_affine_sets_and_convex_sets.ipynb).

**Intuition:**

Each inequality keeps the points on one side of a boundary; stacking several of them leaves a shaded region, and the optimizer may only move inside that region.

**Numerical Example:**

Let
$$
\mathcal{X} = \{(x_1,x_2) : x_1 + x_2 \le 1,\; x_1 \ge 0,\; x_2 \ge 0\}.
$$

Test three points against all three constraints:
$$
(0.2, 0.3):\; 0.5 \le 1,\; 0.2\ge 0,\; 0.3\ge 0 \;\Rightarrow\; \text{feasible},
$$
$$
(0.8, 0.5):\; 1.3 \le 1\; \text{fails} \;\Rightarrow\; \text{infeasible},
$$
$$
(-0.1, 0.4):\; x_1\ge 0\; \text{fails} \;\Rightarrow\; \text{infeasible}.
$$

Only points meeting *every* constraint belong to $\mathcal{X}$; a single violation is enough to exclude a candidate.

```python
import sympy as sp

x_1, x_2 = sp.symbols("x_1 x_2")
constraints = [x_1 + x_2 <= 1, x_1 >= 0, x_2 >= 0]

def is_feasible(point):
    assignment = {x_1: point[0], x_2: point[1]}
    return all(bool(constraint.subs(assignment)) for constraint in constraints)

candidates = [(sp.Rational(1, 5), sp.Rational(3, 10)),
              (sp.Rational(4, 5), sp.Rational(1, 2)),
              (sp.Rational(-1, 10), sp.Rational(2, 5))]

for candidate in candidates:
    print(candidate, "->", "feasible" if is_feasible(candidate) else "infeasible")
```

**Equivalent casadi implementation:**

```python
import casadi as ca

decision_variable = ca.SX.sym("x", 2)
constraint_vector = ca.vertcat(decision_variable[0] + decision_variable[1], decision_variable[0], decision_variable[1])
constraints = ca.Function("g", [decision_variable], [constraint_vector])
lower_bound = [-ca.inf, 0, 0]
upper_bound = [1, ca.inf, ca.inf]

for candidate in ([0.2, 0.3], [0.8, 0.5], [-0.1, 0.4]):
    values = constraints(candidate).full().flatten()
    feasible = all(lower_bound[i] <= values[i] <= upper_bound[i] for i in range(3))
    print(candidate, "->", "feasible" if feasible else "infeasible")
```
