### 1.3.4.2. Infinite Series and Partial Sums

$$
\sum_{k=1}^{\infty} a_k = \lim_{n \to \infty} S_n, \qquad S_n = \sum_{k=1}^{n} a_k ;
\qquad
\sum_{k=0}^{\infty} r^{k} = \frac{1}{1 - r}\ \ (|r| < 1).
$$

**Explanation:**

An infinite series is the limit of its sequence of *partial sums* $S_n$: if the partial sums [converge](./01_sequences_and_convergence.ipynb) to a finite value the series converges to that sum, otherwise it diverges. The geometric series is the prototype — it converges precisely when the ratio satisfies $|r| < 1$, with the closed form $\sum_{k=0}^{\infty} r^k = \tfrac{1}{1-r}$. That formula is exactly the infinite-horizon discounted return $\sum_k \gamma^k$ used throughout reinforcement learning, and a necessary condition for any series to converge is that its terms tend to zero.

**Properties:**
- $n$th-term test: if $a_k \not\to 0$ the series diverges (the converse is false).
- The harmonic series $\sum 1/k$ diverges even though $1/k \to 0$, showing term-decay alone is insufficient.

**Numerical Example:**

Sum the geometric series with $r = \tfrac12$: $\sum_{k=0}^{\infty} \left(\tfrac12\right)^k$. The partial sums are

$$
S_1 = 1, \quad S_2 = 1.5, \quad S_3 = 1.75, \quad S_4 = 1.875, \ldots
$$

approaching the closed-form value $\dfrac{1}{1 - 1/2} = 2$. The remaining "tail" after $S_n$ is $2 - S_n = (1/2)^{n-1} \to 0$, confirming convergence to $2$.

```python
import sympy as sp

k, n = sp.symbols("k n")
ratio = sp.Rational(1, 2)

closed_form = sp.summation(ratio**k, (k, 0, sp.oo))
partial_sums = [sp.summation(ratio**k, (k, 0, upper)) for upper in range(0, 4)]

print("closed form  Σ (1/2)^k =", closed_form)
for upper, value in enumerate(partial_sums, start=1):
    print(f"S_{upper} = {value} = {float(value)}")
```

**Equivalent NumPy implementation:**

Accumulating the geometric terms numerically shows the partial sums climbing toward the closed-form limit $2$.

```python
import numpy as np

ratio = 0.5
terms = ratio ** np.arange(0, 20)
partial_sums = np.cumsum(terms)

print("S_5  =", partial_sums[4])
print("S_20 =", partial_sums[19])
print("closed form 1/(1-r) =", 1 / (1 - ratio))
```
