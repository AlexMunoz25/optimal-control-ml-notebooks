### 1.3.5.2. Partial Derivatives

$$
\frac{\partial f}{\partial x}(x, y) = \lim_{h \to 0} \frac{f(x + h, y) - f(x, y)}{h},
\qquad
\frac{\partial f}{\partial y}(x, y) = \lim_{h \to 0} \frac{f(x, y + h) - f(x, y)}{h}.
$$

**Explanation:**

A partial derivative measures the rate of change of $f$ as **one** input varies while the others are held fixed. Computationally this means differentiating with respect to that variable and treating every other variable as a constant, so each one-variable [differentiation rule](../02_Single_Variable_Differentiation/02_differentiation_rules.ipynb) applies directly. Geometrically $\partial f/\partial x$ is the slope of the surface along the $x$-direction trace. The collection of all partial derivatives is assembled into the [gradient](./03_gradient.ipynb), the central object of multivariable optimization.

**Numerical Example:**

Let $f(x, y) = x^2 y + 3xy^2$. Holding $y$ fixed and differentiating in $x$,

$$
\frac{\partial f}{\partial x} = 2xy + 3y^2 ;
$$

holding $x$ fixed and differentiating in $y$,

$$
\frac{\partial f}{\partial y} = x^2 + 6xy .
$$

At the point $(1, 2)$: $f_x(1,2) = 2(1)(2) + 3(4) = 16$ and $f_y(1,2) = 1 + 6(1)(2) = 13$.

```python
import sympy as sp

x, y = sp.symbols("x y")
f = x**2 * y + 3 * x * y**2

partial_x = sp.diff(f, x)
partial_y = sp.diff(f, y)
evaluation = {x: 1, y: 2}

print("∂f/∂x =", partial_x, " -> at (1,2):", partial_x.subs(evaluation))
print("∂f/∂y =", partial_y, " -> at (1,2):", partial_y.subs(evaluation))
```

**Equivalent CasADi implementation:**

CasADi computes the same partials by differentiating with respect to each component of the input vector.

```python
import casadi as ca

variables = ca.SX.sym("v", 2)
f = variables[0]**2 * variables[1] + 3 * variables[0] * variables[1]**2
partials = ca.gradient(f, variables)
partials_function = ca.Function("partials", [variables], [partials])

print("[∂f/∂x, ∂f/∂y] at (1,2) =", partials_function([1, 2]))
```
