### 1.1.9.13. Vectors and One-Forms

$$
\vec{v} = v^i \vec{e}_i \quad \text{(vector, contravariant)},
\qquad
\tilde{\omega} = \omega_i \vec{e}^{\,i} \quad \text{(one-form, covariant)}
$$

$$
\langle \tilde{\omega},\, \vec{v} \rangle = \omega_i v^i \in \mathbb{R} \quad \text{(invariant scalar pairing)}
$$

**Explanation:**

- In modern terminology, vectors and one-forms are different rank-1 tensors.
- Vectors use contravariant components with basis vectors, while one-forms use covariant components with the dual basis.
- Their pairing $\omega_i v^i$ is invariant and needs no metric because one lower index contracts with one upper index.
- The metric gives an additional map that turns a vector into its associated one-form.

**Connections:**

- This reframes the earlier dual-basis, covariant-component, and contravariant-component notebooks in modern language.
- In ML and optimization, gradients are naturally one-forms; converting them into update vectors depends on the chosen metric or preconditioner.

**Example:**

- For a scalar field $f$, the gradient is the one-form
$$
\tilde{\nabla} f = \frac{\partial f}{\partial x^i}\,\vec{e}^{\,i}.
$$
- Pairing it with a vector $\vec{v} = v^i \vec{e}_i$ gives the directional derivative
$$
\langle \tilde{\nabla} f, \vec{v} \rangle = \frac{\partial f}{\partial x^i} v^i.
$$


```python
import numpy as np

vector_components = np.array([3.0, 1.0, 2.0])
one_form_components = np.array([1.0, 4.0, 2.0])
cartesian_metric = np.eye(3)

scalar_pairing = np.einsum('i,i->', one_form_components, vector_components)
lowered_vector = cartesian_metric @ vector_components
metric_norm_squared = np.einsum(
    'ij,i,j->',
    cartesian_metric,
    vector_components,
    vector_components,
)

print("scalar_pairing =", scalar_pairing)
print("lowered_vector =", lowered_vector)
print("metric_norm_squared =", metric_norm_squared)
```
