### 1.3.4.3. Power Series

$$
\sum_{n=0}^{\infty} c_n (x - a)^{n},
\qquad
\text{radius of convergence } R = \lim_{n\to\infty}\left|\frac{c_n}{c_{n+1}}\right|,
\quad |x - a| < R .
$$

**Explanation:**

A power series is an "infinite polynomial" in $(x - a)$ with coefficients $c_n$, centered at $a$. For each $x$ it is an [infinite series](./02_infinite_series_and_partial_sums.ipynb) of numbers, and the set of $x$ where it converges is an interval of radius $R$ about $a$ — converging for $|x - a| < R$, diverging for $|x - a| > R$, with the endpoints checked separately. Inside that interval the series defines a function that can be differentiated and integrated term by term, which is what makes power series the representation behind [Taylor approximation](./04_taylor_and_maclaurin_series.ipynb).

**Properties:**
- The radius $R$ can be found by the ratio test: the series converges where $\lim_n |c_{n+1}(x-a)^{n+1} / c_n(x-a)^n| < 1$.
- Term-by-term differentiation and integration preserve the radius of convergence.

**Numerical Example:**

The geometric power series $\sum_{n=0}^{\infty} x^n$ has all $c_n = 1$. The ratio test gives

$$
\lim_{n\to\infty}\left|\frac{x^{n+1}}{x^n}\right| = |x| < 1,
$$

so the radius of convergence is $R = 1$ and on $(-1, 1)$ it sums to $\dfrac{1}{1 - x}$. For the exponential series $\sum_{n=0}^{\infty} \dfrac{x^n}{n!}$, the ratio $\dfrac{|x|^{n+1}/(n+1)!}{|x|^n/n!} = \dfrac{|x|}{n+1} \to 0$ for every $x$, so $R = \infty$ — it converges on all of $\mathbb{R}$.

```python
import sympy as sp

x, n = sp.symbols("x n")

geometric_sum = sp.summation(x**n, (n, 0, sp.oo))
geometric_radius = sp.Abs(sp.limit(1 / 1, n, sp.oo))

exponential_ratio = sp.limit((sp.factorial(n)) / sp.factorial(n + 1), n, sp.oo)

print("Σ x^n            =", geometric_sum, " (valid for |x| < 1)")
print("geometric radius R =", geometric_radius)
print("exponential ratio |c_n/c_{n+1}| -> R =", 1 / exponential_ratio if exponential_ratio != 0 else sp.oo)
```
