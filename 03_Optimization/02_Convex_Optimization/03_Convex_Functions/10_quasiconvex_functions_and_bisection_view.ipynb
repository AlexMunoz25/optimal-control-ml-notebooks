### 3.2.3.10. Quasiconvex Functions and Bisection View


$$f(\theta x+(1-\theta)y)\leq\max\{f(x),f(y)\}$$


**Explanation:**

A quasiconvex function is organized by convex sublevel sets rather than by a convex epigraph. This makes level-set feasibility the main tool: fix a candidate value, test whether its sublevel set is feasible, and then adjust the value.

The left side evaluates $f$ at a point between $x$ and $y$. Quasiconvexity requires that value to be no larger than the worse endpoint value, so sublevel sets remain convex even when the function is not convex.

**Properties:**
- [Convex functions](01_convex_function_definition_and_examples.ipynb) have convex [epigraphs](05_sublevel_sets_and_epigraphs.ipynb).
- Quasiconvex functions have convex sublevel sets.
- Epigraph transformations replace objectives by constraints.

**Intuition:**

The figure above shows set membership derived from a function value; the arithmetic checks that membership.

**Numerical Example:**

The figure above shows set membership derived from a function value; the arithmetic checks that membership.

Let $f(x)=x^2$ and $x=2$. First compute
$$
f(2)=2^2=4.
$$

At level $5$,
$$
4\leq5,
$$
so $x=2$ is in the sublevel set. At level $3$,
$$
4\leq3
$$
is false, so $x=2$ is not in that lower sublevel set.


```python
import sympy as sp

function_value = sp.Integer(2) ** 2

print("function_value <= 5 =", function_value <= 5)
print("function_value <= 3 =", function_value <= 3)
```
