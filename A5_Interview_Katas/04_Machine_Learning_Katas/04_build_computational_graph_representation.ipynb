### A5.4.4. Build Computational Graph Representation

**Problem:**

Build a directed acyclic graph (DAG) where nodes represent operations or values and edges represent data flow between them.

**Explanation:**

A computational graph models a computation as a DAG. Each node is either an input value or an operation. Each edge connects an output of one node to an input of another. Building the graph means creating nodes and linking them by recording each node's inputs.

How to build it:

1. Create a `Node` class with a name, an operation type (or `"input"` for leaf nodes), and a list of input nodes.
2. Create a `Graph` class that keeps a list of all nodes.
3. `add_input(name)`: create a node with no inputs and operation type `"input"`.
4. `add_operation(name, op_type, inputs)`: create a node whose inputs are other existing nodes.
5. The graph is now a DAG you can traverse for scheduling, differentiation, or optimization.

**Example:**

Build the graph for `z = (x + y) * y`: input nodes `x`, `y`, an `add` node taking `x` and `y`, and a `mul` node taking the `add` result and `y`.

```python
class Node:
    def __init__(self, name, op_type, inputs=None):
        self.name = name
        self.op_type = op_type
        self.inputs = inputs or []

    def __repr__(self):
        input_names = [node.name for node in self.inputs]
        return f"Node({self.name}, op={self.op_type}, inputs={input_names})"


class Graph:
    def __init__(self):
        self.nodes = []

    def add_input(self, name):
        node = Node(name, "input")
        self.nodes.append(node)
        return node

    def add_operation(self, name, op_type, inputs):
        node = Node(name, op_type, inputs)
        self.nodes.append(node)
        return node


graph = Graph()

x = graph.add_input("x")
y = graph.add_input("y")
add_node = graph.add_operation("add_0", "add", [x, y])
mul_node = graph.add_operation("mul_0", "mul", [add_node, y])

print("Computational graph for z = (x + y) * y:")
for node in graph.nodes:
    print(f"  {node}")
```
