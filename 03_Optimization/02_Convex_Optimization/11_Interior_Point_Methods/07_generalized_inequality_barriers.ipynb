### 3.2.11.7. Generalized Inequality Barriers


$$\phi(X)=-\log\det X$$


**Explanation:**

[Generalized inequality](../02_Convex_Sets/09_generalized_inequalities_and_proper_cones.ipynb) barriers use [cone-specific](../02_Convex_Sets/03_cones_and_conic_combinations.ipynb) logarithmic barriers. For the [positive semidefinite cone](../02_Convex_Sets/06_positive_semidefinite_cone.ipynb), the standard barrier is negative log [determinant](../../../01_Foundations/01_Linear_Algebra/03_Matrix/10_matrix_determinant.ipynb), finite only on [positive definite matrices](../../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/17_positive_definite_matrix.ipynb).

The variable $X$ is a positive definite matrix. The barrier $-\log\det X$ uses the [determinant](../../../01_Foundations/01_Linear_Algebra/03_Matrix/10_matrix_determinant.ipynb) to approach infinity as $X$ approaches the boundary of the positive semidefinite cone.

**Properties:**
- The determinant must be positive.
- Approaching singularity sends the barrier to infinity.
- The log-det barrier encodes positive-definite cone geometry.

**Intuition:**

The example evaluates a log-det barrier for a positive definite matrix.


**Numerical Example:**

Let
$$
X=\begin{bmatrix}2&1\\1&2\end{bmatrix}.
$$

The determinant is
$$
\det X=2\cdot2-1\cdot1=3.
$$

The log-det barrier is
$$
-\log\det X=-\log3.
$$

The value is finite because $X$ is positive definite.


```python
import sympy as sp

matrix = sp.Matrix([[2, 1], [1, 2]])
determinant = matrix.det()
barrier_value = -sp.log(determinant)

determinant, barrier_value
```
