### 1.1.7.20. QR Decomposition


$$
A = Q R
$$


**Explanation:**

The QR decomposition factors a matrix into an orthogonal matrix and an upper-triangular matrix. It is closely tied to [Gram-Schmidt orthogonalization](./16_gram_schmidt_orthogonalization.ipynb), which produces the orthonormal columns of $Q$.

**Example:**

The columns of $Q$ are the orthonormalized columns of $A$, while $R$ stores the coefficients needed to reconstruct the original matrix.


```python
import numpy as np

matrix_a = np.array([[1.0, 1.0], [1.0, 0.0], [0.0, 1.0]])
matrix_q, matrix_r = np.linalg.qr(matrix_a)

print("matrix_q =")
print(np.round(matrix_q, 4))
print("matrix_r =")
print(np.round(matrix_r, 4))
```
