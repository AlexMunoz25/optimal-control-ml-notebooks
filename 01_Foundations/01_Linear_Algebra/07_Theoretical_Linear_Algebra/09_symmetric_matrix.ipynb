### 1.1.7.9. Symmetric Matrix


$$
A^{\mathsf{T}} = A
$$


**Explanation:**

A symmetric matrix equals its transpose. Its main structural consequences are real eigenvalues and orthogonal eigenvectors, while the stronger positivity condition is treated separately in [positive definite matrices](./10_positive_definite_matrix.ipynb).

**Example:**

The matrix

$$
\begin{bmatrix}
2 & 1 \\
1 & 3
\end{bmatrix}
$$

is symmetric because the off-diagonal entries match.


```python
import sympy as sp

matrix_a = sp.Matrix([[2, 1], [1, 3]])

print("is_symmetric =", matrix_a == matrix_a.T)
print("eigenvalues =", list(matrix_a.eigenvals().keys()))
```
