### 1.1.7.11. Eigenvalues and Invertibility


$$
0 \notin \operatorname{eig}(A)
\Longleftrightarrow
A \text{ is invertible}
$$


**Explanation:**

A square matrix is invertible exactly when zero is not one of its [eigenvalues](./01_eigenvalue.ipynb). This criterion follows from the determinant product formula and sharpens one part of the broader [invertible matrix theorem](../06_Linear_Transformations/14_invertible_matrix_theorem.ipynb).

**Example:**

If a matrix has eigenvalues $5$ and $10$, then it is invertible because neither eigenvalue is zero. If one eigenvalue were $0$, the matrix would have a nontrivial kernel and could not be inverted.


```python
import sympy as sp

matrix_a = sp.Matrix([[9, -2], [-2, 6]])
eigenvalues = list(matrix_a.eigenvals().keys())
is_invertible = 0 not in eigenvalues

print("eigenvalues =", eigenvalues)
print("is_invertible =", is_invertible)
```
