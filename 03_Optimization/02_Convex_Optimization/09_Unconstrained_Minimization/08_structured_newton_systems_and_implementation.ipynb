### 3.2.9.8. Structured Newton Systems and Implementation


$$\nabla^2f(x)\Delta x=-\nabla f(x)$$


**Explanation:**

Structured Newton implementations exploit linear-system structure in Hessian or KKT matrices. The mathematical step is still a Newton solve, but sparse or block structure makes the computation scalable while the residual equations preserve the same optimality meaning.

This is the linear system for the Newton step $\Delta x$. The Hessian matrix multiplies the unknown step, and the right-hand side is the negative gradient, so solving the [matrix equation](../../../01_Foundations/01_Linear_Algebra/04_Computational_Linear_Algebra/06_matrix_equations.ipynb) gives the direction.

**Properties:**
- The coefficient matrix represents local curvature and constraints.
- The right-hand side represents a residual to reduce.
- Structured solves are essential for large convex optimization problems.

**Intuition:**

The example solves a small structured linear system exactly.


**Numerical Example:**

Solve
$$
\begin{bmatrix}4&1\\1&3\end{bmatrix}
\begin{bmatrix}\Delta z_1\\\Delta z_2\end{bmatrix}
=
\begin{bmatrix}1\\2\end{bmatrix}.
$$

The [determinant](../../../01_Foundations/01_Linear_Algebra/03_Matrix/10_matrix_determinant.ipynb) is
$$
4\cdot3-1\cdot1=11.
$$

Thus
$$
\Delta z_1=\frac{3\cdot1-1\cdot2}{11}=\frac1{11},
$$
and
$$
\Delta z_2=\frac{-1\cdot1+4\cdot2}{11}=\frac7{11}.
$$


```python
import sympy as sp

coefficient_matrix = sp.Matrix([[4, 1], [1, 3]])
right_side = sp.Matrix([1, 2])
solution = coefficient_matrix.LUsolve(right_side)

print("solution =", solution)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

coefficient_matrix = ca.DM([[4, 1], [1, 3]])
right_side = ca.DM([1, 2])
solution = ca.solve(coefficient_matrix, right_side)

print("solution =", solution)
```
