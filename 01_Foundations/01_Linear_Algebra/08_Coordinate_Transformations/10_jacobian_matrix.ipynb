### 1.1.8.10. Jacobian Matrix


$$
J^{i}_{\;j} = \frac{\partial x'^{\,i}}{\partial x^j}
$$


**Explanation:**

The Jacobian matrix records the local linear action of a coordinate transformation. Fleisch uses it to express how contravariant and covariant components transform once the coordinate map is known. This notebook depends on the notation introduced in [Einstein summation convention](./09_einstein_summation_convention.ipynb).

**Example:**

For a polar-to-Cartesian change of coordinates, the Jacobian entries are the partial derivatives of $x$ and $y$ with respect to $r$ and $\theta$.


```python
import sympy as sp

symbol_r, symbol_theta = sp.symbols("symbol_r symbol_theta")
jacobian_matrix = sp.Matrix(
    [
        [sp.diff(symbol_r * sp.cos(symbol_theta), symbol_r), sp.diff(symbol_r * sp.cos(symbol_theta), symbol_theta)],
        [sp.diff(symbol_r * sp.sin(symbol_theta), symbol_r), sp.diff(symbol_r * sp.sin(symbol_theta), symbol_theta)],
    ]
)

print("jacobian_matrix =")
sp.pprint(jacobian_matrix)
```

**References:**
