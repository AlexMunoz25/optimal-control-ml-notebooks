### 1.1.7.30. Spectral Radius


$$
\rho(A) = \max_{\lambda_i \in \sigma(A)} |\lambda_i|
$$


**Explanation:**

The spectral radius is the largest magnitude among the [eigenvalues](./08_eigenvalue.ipynb) of a square matrix. It measures the spectral growth factor that appears when studying repeated matrix multiplication, stability, and convergence of iterative methods. Because it uses $|\lambda_i|$, real and complex eigenvalues are compared by distance from zero.

**Example:**

For

$$
A =
\begin{bmatrix}
2 & 0 & 0 \\
0 & -3 & 0 \\
0 & 0 & \frac{1}{2}
\end{bmatrix},
$$

the eigenvalues are $2$, $-3$, and $\frac{1}{2}$, so

$$
\rho(A) = \max\left\{2, 3, \frac{1}{2}\right\} = 3.
$$


```python
import sympy as sp

matrix_a = sp.Matrix([[2, 0, 0], [0, -3, 0], [0, 0, sp.Rational(1, 2)]])
eigenvalues = list(matrix_a.eigenvals().keys())
spectral_radius = max(abs(eigenvalue) for eigenvalue in eigenvalues)

print("eigenvalues =", eigenvalues)
print("spectral_radius =", spectral_radius)
```
