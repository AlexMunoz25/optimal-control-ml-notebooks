### 3.2.1.2. Least Squares, Linear Programming, and Convexity


$$\min_x\|Ax-b\|_2^2,\qquad \min_x c^Tx\ \text{subject to}\ Gx\preceq h$$


**Explanation:**

Least squares and linear programming are two basic convex models with different geometry. Least squares uses a convex quadratic residual objective, while a linear program uses a linear objective over a polyhedron. Both make feasibility and optimality explicit enough to compute and certify in small examples.

In the first model, the [matrix](../../../01_Foundations/01_Linear_Algebra/03_Matrix/01_matrix_definition.ipynb) $A$ maps the decision vector $x$ to predictions $Ax$, and the squared [norm](../../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb) measures the residual $Ax-b$. In the second model, $c^Tx$ is a linear objective and $Gx\preceq h$ is a vector inequality constraint.

**Properties:**
- Residuals measure mismatch between a model and data.
- Squaring residuals gives a smooth convex objective.
- The derivative or normal equations give the optimality condition.

**Intuition:**

The computation expands the residual objective so the convex quadratic structure is visible before solving.


**Numerical Example:**

Let $r_1(x)=x-1$ and $r_2(x)=2x-3$. The least-squares objective is
$$
f(x)=(x-1)^2+(2x-3)^2.
$$

Expand it:
$$
f(x)=x^2-2x+1+4x^2-12x+9=5x^2-14x+10.
$$

Differentiate and set the derivative to zero:
$$
f'(x)=10x-14=0.
$$

Thus
$$
x^\star=\frac{14}{10}=\frac75.
$$

The minimum value is
$$
f(7/5)=(2/5)^2+(-1/5)^2=\frac15.
$$


```python
import sympy as sp

variable = sp.symbols('x')
objective = (variable - 1) ** 2 + (2 * variable - 3) ** 2
optimizer = sp.solve(sp.diff(objective, variable), variable)[0]
minimum_value = sp.simplify(objective.subs(variable, optimizer))

optimizer, minimum_value
```
