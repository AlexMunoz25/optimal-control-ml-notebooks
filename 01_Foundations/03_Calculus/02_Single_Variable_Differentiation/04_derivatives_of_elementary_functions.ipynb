### 1.3.2.4. Derivatives of Elementary Functions

$$
\frac{d}{dx}e^{x} = e^{x}, \qquad
\frac{d}{dx}\ln x = \frac{1}{x}, \qquad
\frac{d}{dx}b^{x} = b^{x}\ln b, \qquad
\frac{d}{dx}\log_b x = \frac{1}{x \ln b},
$$

$$
\frac{d}{dx}\sin x = \cos x, \qquad
\frac{d}{dx}\cos x = -\sin x, \qquad
\frac{d}{dx}\tan x = \sec^{2} x .
$$

**Explanation:**

These are the derivatives of the standard building-block functions, memorized once and then combined with the product, quotient, and [chain rules](./03_chain_rule.ipynb) to differentiate any elementary expression. The natural exponential is the unique function equal to its own derivative, which is why $e^x$ pervades growth models and the natural logarithm's derivative $1/x$ underlies log-likelihoods. The trigonometric derivatives recur in oscillatory dynamics and in the periodic features used in modelling.

**Properties:**
- General bases reduce to the natural ones: $b^x = e^{x\ln b}$ and $\log_b x = \ln x / \ln b$.
- Each formula is derived from the limit definition together with the two special limits $\lim_{x\to 0}\frac{\sin x}{x} = 1$ and $\lim_{h\to 0}\frac{e^h - 1}{h} = 1$.

**Numerical Example:**

Differentiate $f(x) = e^{x}\sin x$ by combining the exponential and sine derivatives through the product rule:

$$
f'(x) = \frac{d}{dx}(e^x)\,\sin x + e^x\,\frac{d}{dx}(\sin x) = e^x \sin x + e^x \cos x = e^x(\sin x + \cos x).
$$

Evaluate the elementary derivatives at sample points: $\frac{d}{dx}e^x\big|_{x=0} = e^0 = 1$, $\frac{d}{dx}\ln x\big|_{x=2} = \tfrac12$, and $\frac{d}{dx}\sin x\big|_{x=0} = \cos 0 = 1$.

```python
import sympy as sp

x = sp.symbols("x", positive=True)
b = sp.symbols("b", positive=True)

elementary = {
    "e^x": sp.exp(x),
    "ln x": sp.log(x),
    "b^x": b**x,
    "log_b x": sp.log(x, b),
    "sin x": sp.sin(x),
    "cos x": sp.cos(x),
    "tan x": sp.tan(x),
}
derivatives = {name: sp.simplify(sp.diff(expression, x)) for name, expression in elementary.items()}

product_example = sp.exp(x) * sp.sin(x)
product_derivative = sp.simplify(sp.diff(product_example, x))

for name, derivative in derivatives.items():
    print(f"d/dx {name:8} = {derivative}")
print("d/dx (e^x sin x) =", sp.factor(product_derivative))
```

**Equivalent CasADi implementation:**

CasADi supplies the same elementary derivatives from its built-in operations.

```python
import casadi as ca

x = ca.SX.sym("x")
expressions = {"e^x": ca.exp(x), "ln x": ca.log(x), "sin x": ca.sin(x), "cos x": ca.cos(x)}

for name, expression in expressions.items():
    print(f"d/dx {name:6} =", ca.simplify(ca.gradient(expression, x)))
```
