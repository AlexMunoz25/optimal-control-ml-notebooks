### 3.1.30. Practical Guidelines


$$
\tilde f(y)=f(Sy),
\qquad
|x| \approx \sqrt{x^2+\epsilon},
\qquad
\min_x f(x).
$$


**Explanation:**

Successful optimization is not only a matter of choosing an algorithm. The formulation, scaling, derivative availability, stopping criteria, and validation of the computed result all affect the outcome. Poor scaling can make a simple problem numerically difficult, while a carefully scaled model can make a basic method surprisingly effective. Smoothing can make a nonsmooth model compatible with differentiable methods, but it also changes the problem, so it must be used deliberately.

The display collects three practical modeling moves. The substitution $\tilde f(y)=f(Sy)$ rescales variables, the square-root expression smooths the absolute value, and $\min_x f(x)$ reminds the reader that these changes are judged by how they affect the original minimization problem.

For an engineering student, these guidelines connect the theory back to modeling practice. A method should be selected according to the structure of the cost, the cost of derivative evaluations, the dimension of the problem, and the accuracy actually needed. Validation then checks whether the computed point satisfies the relevant optimality conditions and whether the result makes sense for the original engineering problem.

**Intuition:**

The figure shows smoothing of a nondifferentiable term, a common practical device when derivative-based algorithms are desired. The code cell compares $|x|$ with a smooth approximation $\sqrt{x^2+\epsilon}$ at a few points. The approximation is close away from the origin, but near the origin it deliberately rounds the sharp corner, illustrating both the benefit and the modeling change.

<img src="../../Figures/030101_berts_fig_1_10_1_smoothing_nondifferentiable_terms.jpeg" alt="Bertsekas Figure 1.10.1: smoothing nondifferentiable terms" style="width: 60%; height: auto;">


**Numerical Example:**

The smoothing figure replaces a sharp corner by a differentiable curve. Compare
$$
|x|
\qquad\text{with}\qquad
\sqrt{x^2+\epsilon},
\qquad \epsilon=0.01.
$$

At $x=1$,
$$
\sqrt{1^2+0.01}=\sqrt{1.01}\approx 1.00499,
$$
which is close to $|1|=1$. At $x=0.1$,
$$
\sqrt{0.1^2+0.01}=\sqrt{0.02}\approx 0.14142,
$$
which is larger than $|0.1|=0.1$.

At the corner $x=0$,
$$
\sqrt{0^2+0.01}=0.1,
\qquad |0|=0.
$$

Thus smoothing stays close away from the corner but deliberately changes the objective near the nondifferentiable point.


```python
import sympy as sp

values = sp.Matrix([-1, sp.Rational(-1, 10), 0, sp.Rational(1, 10), 1])
smoothing_parameter = sp.Rational(1, 100)
absolute_values = values.applyfunc(sp.Abs)
smoothed_values = values.applyfunc(lambda value: sp.sqrt(value ** 2 + smoothing_parameter))

print("absolute_values ="); sp.pprint(absolute_values.T)
print("smoothed_values ="); sp.pprint(smoothed_values.T)
```
