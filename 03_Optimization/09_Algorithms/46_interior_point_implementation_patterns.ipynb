### 3.9.46. Interior-Point Implementation Patterns

$$K\Delta z=r$$


**Explanation:**

Interior-point implementations are dominated by structured linear solves. The mathematical residual equations define the right-hand side, and sparse or block structure makes the solve practical at scale.

The matrix $K$ is the linearized KKT or Newton system matrix, $\Delta z$ is the stacked primal-dual step, and $r$ is the residual right-hand side. Solving this [matrix equation](../../01_Foundations/01_Linear_Algebra/04_Computational_Linear_Algebra/06_matrix_equations.ipynb) is the main computation in each interior-point iteration.

**Properties:**
- The coefficient matrix represents local curvature and constraints.
- The right-hand side represents residual reduction.
- Structured linear algebra is essential for large conic and barrier problems.

**Intuition:**

The example solves the small linear algebra core exactly.


**Numerical Example:**

Solve
$$
K\Delta z=r,
\qquad
K=\begin{bmatrix}4&1\\1&3\end{bmatrix},
\qquad
r=\begin{bmatrix}1\\2\end{bmatrix}.
$$

The [determinant](../../01_Foundations/01_Linear_Algebra/03_Matrix/10_matrix_determinant.ipynb) is
$$
4\cdot3-1\cdot1=11.
$$

Therefore
$$
\Delta z_1=\frac{3\cdot1-1\cdot2}{11}=\frac1{11},
\qquad
\Delta z_2=\frac{-1\cdot1+4\cdot2}{11}=\frac7{11}.
$$


```python
import sympy as sp

coefficient_matrix = sp.Matrix([[4, 1], [1, 3]])
right_side = sp.Matrix([1, 2])
solution = coefficient_matrix.LUsolve(right_side)

print("solution =", solution)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

coefficient_matrix = ca.DM([[4, 1], [1, 3]])
right_side = ca.DM([1, 2])
solution = ca.solve(coefficient_matrix, right_side)

print("solution =", solution)
```
