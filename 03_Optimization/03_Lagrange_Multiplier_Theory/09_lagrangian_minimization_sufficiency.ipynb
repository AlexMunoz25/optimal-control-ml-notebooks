### 3.3.9. Lagrangian Minimization Sufficiency

$$
x^*\in \arg\min_{x\in X}\left(f(x)+\sum_{j=1}^r\mu_j^*g_j(x)\right),
\qquad
\mu^*\ge 0,
\qquad
\mu_j^*g_j(x^*)=0
$$

**Explanation:**

A point is globally optimal if it minimizes the Lagrangian over the simple set $X$, has nonnegative inequality multipliers, and satisfies complementary slackness. This is a sufficiency statement: it proves optimality directly from the Lagrangian lower-bound argument, without needing second derivatives. When the objective and constraints are convex and differentiable, the minimization condition is often verified through first-order stationarity.

**Numerical Example:**

Consider

$$
\min x^2 \quad \text{subject to} \quad 1-x\le 0.
$$

The constraint is equivalent to $x\ge 1$. The Lagrangian is

$$
L(x,\mu)=x^2+\mu(1-x).
$$

Try the feasible point $x^*=1$. The active constraint gives

$$
g(1)=1-1=0.
$$

Stationarity of the Lagrangian in $x$ gives

$$
\frac{dL}{dx}=2x-\mu.
$$

At $x^*=1$,

$$
2-\mu=0 \quad \Rightarrow \quad \mu^*=2.
$$

The multiplier is nonnegative, and complementary slackness holds:

$$
\mu^*g(x^*)=2\cdot 0=0.
$$

With $\mu^*=2$,

$$
L(x,2)=x^2+2(1-x)=x^2-2x+2=(x-1)^2+1.
$$

This is minimized at $x=1$, so the Lagrangian minimization sufficiency condition proves that $x^*=1$ is a global minimum.

```python
import sympy as sp

x, mu = sp.symbols('x mu')

cost = x**2
constraint = 1 - x
lagrangian = cost + mu * constraint
candidate = {x: 1}
mu_solution = sp.solve(sp.diff(lagrangian, x).subs(candidate), [mu], dict=True)[0]
lagrangian_at_multiplier = sp.expand(lagrangian.subs(mu_solution))

result = {
    'mu_solution': mu_solution,
    'lagrangian_at_multiplier': lagrangian_at_multiplier,
    'completed_square': sp.factor(lagrangian_at_multiplier - 1) + 1,
    'complementary_slackness': (mu * constraint).subs(candidate).subs(mu_solution),
}
result
```
