### 3.2.8.1. Projection on Convex Sets


$$P_C(x_0)=\arg\min_{x\in C}\|x-x_0\|_2$$


**Explanation:**

Projection maps a point to the closest point in a [convex set](../02_Convex_Sets/01_affine_sets_and_convex_sets.ipynb). The squared-[distance](../../../01_Foundations/01_Linear_Algebra/05_Geometrical_Aspects_of_Linear_Algebra/04_distance_between_points.ipynb) objective is convex, and a closed convex set gives a unique Euclidean projection. Feasibility is membership in the set, while optimality is the closest-point condition.

The point $x_0$ is projected onto the set $C$ by searching over $x\in C$. The expression chooses the feasible point with smallest Euclidean [distance](../../../01_Foundations/01_Linear_Algebra/05_Geometrical_Aspects_of_Linear_Algebra/04_distance_between_points.ipynb) from $x_0$.

**Properties:**
- The projection point is feasible.
- The distance is minimized over the set.
- The projection residual can define a separating hyperplane.

**Intuition:**

The figure above shows closest-point geometry; the calculation projects a scalar point onto an interval.

<img src="../../../Figures/03020801_bv_fig_8_1_projection_on_convex_set.jpeg" alt="Boyd and Vandenberghe Figure 8.1: projection of a point on a convex set" style="width: 60%; height: auto;">


**Numerical Example:**

The figure above shows closest-point geometry; this calculation gives a one-dimensional version.

Project $x_0=3$ onto
$$
C=[-1,1].
$$

The closest feasible point is the right endpoint:
$$
P_C(3)=1.
$$

The distance is
$$
|3-1|=2.
$$

Thus the projection is $1$ and the distance to the set is $2$.


```python
import sympy as sp

point = sp.Integer(3)
projection = sp.Integer(1)
distance = abs(point - projection)

projection, distance
```
