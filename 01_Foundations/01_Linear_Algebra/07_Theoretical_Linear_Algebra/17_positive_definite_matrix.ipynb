### 1.1.7.17. Positive Definite Matrix


$$
\mathbf{v}^{\mathsf{T}} P \mathbf{v} > 0
\qquad
\text{for all } \mathbf{v} \neq \mathbf{0}
$$


**Explanation:**

A positive definite matrix produces a strictly positive quadratic form for every nonzero vector. Although many important examples are [symmetric matrices](./09_symmetric_matrix.ipynb), positivity is the defining idea in its own right. That same condition makes [Cholesky decomposition](./19_cholesky_decomposition.ipynb) possible and encodes positive energy or curvature in optimization and control.

**Example:**

For

$$
P =
\begin{bmatrix}
2 & 0 \\
0 & 3
\end{bmatrix},
$$

the quadratic form is $2 x_1^2 + 3 x_2^2$, which is positive for every nonzero vector.


```python
import sympy as sp

matrix_p = sp.Matrix([[2, 0], [0, 3]])
vector_v = sp.Matrix([1, 2])
quadratic_value = sp.expand((vector_v.T * matrix_p * vector_v)[0])

print("quadratic_value =", quadratic_value)
print("eigenvalues =", list(matrix_p.eigenvals().keys()))
```
