### 1.1.7.19. Cholesky Decomposition


$$
A = L L^{\mathsf{T}}
$$


**Explanation:**

Cholesky decomposition is the symmetric positive-definite specialization of triangular factorization. A positive definite matrix factors as a lower-triangular matrix times its transpose, which is why the condition in [positive definite matrices](./10_positive_definite_matrix.ipynb) is essential.

**Example:**

For a symmetric positive definite matrix, Cholesky uses half the work of a general LU factorization.


```python
import sympy as sp

matrix_a = sp.Matrix([[4, 2], [2, 5]])
matrix_l = matrix_a.cholesky()

print("matrix_l =")
sp.pprint(matrix_l)
print("reconstruction =")
sp.pprint(matrix_l * matrix_l.T)
```
