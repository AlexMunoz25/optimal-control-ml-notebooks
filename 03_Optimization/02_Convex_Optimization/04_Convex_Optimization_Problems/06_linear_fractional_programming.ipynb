### 3.2.4.6. Linear-Fractional Programming


$$f_0(x)=\frac{c^Tx+d}{e^Tx+f},\qquad e^Tx+f>0$$


**Explanation:**

Linear-fractional objectives are ratios of affine functions on domains where the denominator is positive. A perspective-style change of variables normalizes by the denominator and converts the ratio into a linear expression in lifted variables.

The numerator and denominator are affine functions of $x$. The condition $e^Tx+f>0$ defines the domain on which the ratio is meaningful and keeps the denominator sign fixed.

**Properties:**
- The denominator must be positive.
- The lifted variables preserve objective values.
- The transformation turns a fractional expression into an affine one.

**Intuition:**

The example evaluates both the original ratio and the lifted expression at the same point.


**Numerical Example:**

Evaluate
$$
f(x)=\frac{2x+1}{x+3}
$$
at $x=2$. The denominator is
$$
2+3=5>0.
$$

The original value is
$$
f(2)=\frac{2\cdot2+1}{2+3}=\frac55=1.
$$

Define
$$
y=\frac{x}{x+3}=\frac25,
\qquad
s=\frac1{x+3}=\frac15.
$$

Then the lifted expression gives
$$
2y+s=2\cdot\frac25+\frac15=1.
$$


```python
import sympy as sp

point = sp.Integer(2)
denominator = point + 3
original_value = (2 * point + 1) / denominator
transformed_variable = point / denominator
transformed_scale = 1 / denominator
transformed_value = 2 * transformed_variable + transformed_scale

original_value, transformed_value
```
