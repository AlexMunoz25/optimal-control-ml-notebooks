### 3.3.7. Second-Order Cone Programming


$$\|A_ix+b_i\|_2\leq c_i^Tx+d_i$$


**Explanation:**

A second-order cone constraint bounds a [norm](../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb) by a scalar component. SOCPs collect affine equalities and inequalities of this cone form, which covers many robust, quadratic, and norm-constrained models.

The left side is a Euclidean norm of an affine expression in $x$. The right side is another affine expression, so the constraint says that a [norm](../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb) must fit under a scalar affine bound.

**Properties:**
- Norm balls are convex by the triangle inequality.
- Norm [cones](../02_Convex_Analysis/03_cones_and_conic_combinations.ipynb) contain pairs whose scalar component dominates the norm.
- Second-order cone constraints use this pattern.

**Intuition:**

The example uses a three-four-five triangle, so the norm is exact.


**Numerical Example:**

Let
$$
u=\begin{bmatrix}3\\4\end{bmatrix},\quad t=5.
$$

Compute the norm:
$$
\|u\|_2=\sqrt{3^2+4^2}=\sqrt{9+16}=\sqrt{25}=5.
$$

The cone test is
$$
\|u\|_2\leq t.
$$

Substitution gives
$$
5\leq5,
$$
so the point is feasible and lies on the cone boundary.


```python
import sympy as sp

vector = sp.Matrix([3, 4])
height = sp.Integer(5)
norm_value = sp.sqrt((vector.T * vector)[0])

print("norm_value =", norm_value)
print("norm_value <= height =", norm_value <= height)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

vector = ca.DM([3, 4])
height = 5
norm_value = ca.norm_2(vector)

print("norm_value =", norm_value)
print("norm_value <= height =", bool(norm_value <= height))
```

**Feasible-set geometry:**

A second-order cone constraint $\|u\|_2\le t$ carves out the cone whose circular cross-sections grow linearly with the scalar bound $t$. The cell below rebuilds this cone, with the apex at the origin and a flat disk closing it at $t=1$.

<img src="../../Figures/030307_socp_second_order_cone.png" alt="Second-order cone feasible set" style="width: 46%; height: auto;">

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LightSource

height = np.linspace(0, 1, 40)
angle = np.linspace(0, 2 * np.pi, 80)
height_grid, angle_grid = np.meshgrid(height, angle)
cone_x = height_grid * np.cos(angle_grid)
cone_y = height_grid * np.sin(angle_grid)
cone_z = height_grid

disk_radius = np.linspace(0, 1, 30)
disk_radius_grid, disk_angle_grid = np.meshgrid(disk_radius, angle)
top_x = disk_radius_grid * np.cos(disk_angle_grid)
top_y = disk_radius_grid * np.sin(disk_angle_grid)
top_z = np.ones_like(top_x)

light = LightSource(azdeg=200, altdeg=40)
figure = plt.figure(figsize=(5, 5))
axis = figure.add_subplot(111, projection="3d")
axis.plot_surface(cone_x, cone_y, cone_z, color="#b56f24", lightsource=light, linewidth=0, antialiased=True, shade=True)
axis.plot_surface(top_x, top_y, top_z, color="#e89a3c", linewidth=0, shade=False)
axis.set_xlim(-1, 1)
axis.set_ylim(-1, 1)
axis.set_zlim(0, 1.05)
axis.view_init(elev=16, azim=-60)
axis.set_box_aspect((1, 1, 0.8))
for pane in (axis.xaxis.pane, axis.yaxis.pane, axis.zaxis.pane):
    pane.set_facecolor((1, 1, 1, 0))
    pane.set_edgecolor("#cfcfcf")
axis.set_xticklabels([])
axis.set_yticklabels([])
axis.set_zticklabels([])
axis.tick_params(length=0)
figure.savefig("../../Figures/030307_socp_second_order_cone.png", dpi=150, bbox_inches="tight")
plt.show()
```
