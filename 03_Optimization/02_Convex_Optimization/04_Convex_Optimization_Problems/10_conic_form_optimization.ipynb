### 3.2.4.10. Conic Form Optimization


$$\min_x c^Tx\quad\text{subject to}\quad Fx+g\preceq_K0$$


**Explanation:**

Conic form optimization writes the constraint residual as membership in a cone. Linear programs, second-order cone programs, and semidefinite programs fit the same template once the appropriate cone is chosen.

The objective $c^Tx$ is linear in the decision vector. The constraint $Fx+g\preceq_K0$ means the affine residual is ordered by the cone $K$, so feasibility is cone membership after an [affine transformation](../../../01_Foundations/01_Linear_Algebra/06_Linear_Transformations/08_affine_transformation.ipynb).

**Properties:**
- The cone defines the meaning of inequality.
- For the nonnegative orthant, every component must be nonnegative.
- This order supports vector optimization and conic duality.

**Intuition:**

The example checks cone membership by inspecting each component of a difference vector.


**Numerical Example:**

Let
$$
x=\begin{bmatrix}1\\2\end{bmatrix},\quad
y=\begin{bmatrix}3\\5\end{bmatrix},\quad K=\mathbf{R}_+^2.
$$

The order test is $x\preceq_K y$ if $y-x\in K$. Compute
$$
y-x=\begin{bmatrix}3-1\\5-2\end{bmatrix}=\begin{bmatrix}2\\3\end{bmatrix}.
$$

Both components are nonnegative, so $x\preceq_K y$.


```python
import sympy as sp

point_x = sp.Matrix([1, 2])
point_y = sp.Matrix([3, 5])
difference = point_y - point_x
nonnegative = [entry >= 0 for entry in difference]

difference, nonnegative
```
