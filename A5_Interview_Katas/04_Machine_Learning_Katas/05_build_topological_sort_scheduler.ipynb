### A5.4.5. Build Topological Sort Scheduler

**Problem:**

Build a scheduler that takes a computational graph and returns a valid execution order — every node is scheduled only after all its inputs have been scheduled.

**Explanation:**

Topological sort produces a linear ordering of nodes in a DAG such that every node appears after all its dependencies. This is the execution schedule for a computational graph.

How to build it (Kahn's algorithm):

1. Compute the in-degree of each node (how many inputs it has).
2. Put all nodes with in-degree 0 (inputs with no dependencies) into a queue.
3. While the queue is not empty: pop a node, add it to the schedule, and for each node that uses this node as an input, decrement its in-degree. If a node's in-degree reaches 0, add it to the queue.
4. The schedule is the list of nodes in the order they were popped.

**Example:**

For `z = (x + y) * y`, topological sort yields `[x, y, add, mul]` — inputs first, then operations in dependency order.

```python
from collections import deque


class Node:
    def __init__(self, name, inputs=None):
        self.name = name
        self.inputs = inputs or []

    def __repr__(self):
        return self.name


def topological_sort(nodes):
    consumers = {node: [] for node in nodes}
    in_degree = {node: len(node.inputs) for node in nodes}

    for node in nodes:
        for input_node in node.inputs:
            consumers[input_node].append(node)

    ready_nodes = deque(
        node for node in nodes
        if in_degree[node] == 0
    )

    schedule = []
    while ready_nodes:
        current = ready_nodes.popleft()
        schedule.append(current)
        for consumer in consumers[current]:
            in_degree[consumer] -= 1
            if in_degree[consumer] == 0:
                ready_nodes.append(consumer)

    return schedule


x = Node("x")
y = Node("y")
add_node = Node("add", [x, y])
mul_node = Node("mul", [add_node, y])
neg_node = Node("neg", [mul_node])

all_nodes = [neg_node, mul_node, x, add_node, y]
schedule = topological_sort(all_nodes)

print(f"Execution order: {schedule}")
```
