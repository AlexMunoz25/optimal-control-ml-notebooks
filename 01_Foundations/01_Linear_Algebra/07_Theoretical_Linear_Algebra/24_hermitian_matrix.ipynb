### 1.1.7.24. Hermitian Matrix


$$
H^{\mathsf{H}} = H
$$


**Explanation:**

A Hermitian matrix is the complex analogue of a [symmetric matrix](./09_symmetric_matrix.ipynb). Its key structural features are real eigenvalues and orthogonal eigenvectors with respect to the complex inner product.

**Example:**

Let

$$
H=\begin{bmatrix}2&1+i\\1-i&3\end{bmatrix}.
$$

The conjugate transpose swaps the off-diagonal entries and conjugates them:

$$
H^*=\begin{bmatrix}2&1+i\\1-i&3\end{bmatrix}=H.
$$

The determinant part of the characteristic polynomial is

$$
(2-\lambda)(3-\lambda)-(1+i)(1-i).
$$

Since $(1+i)(1-i)=2$, this becomes

$$
(2-\lambda)(3-\lambda)-2=\lambda^2-5\lambda+4=(\lambda-1)(\lambda-4).
$$

Thus the eigenvalues are real: $1$ and $4$.


```python
import sympy as sp

matrix_h = sp.Matrix([[2, 1 + sp.I], [1 - sp.I, 3]])

print("is_hermitian =", matrix_h == matrix_h.conjugate().T)
print("eigenvalues =", list(matrix_h.eigenvals().keys()))
```
