### 3.2.8.6. Chebyshev, Analytic, and Ellipsoid Centers


$$\max_{x_c,r}r$$


**Explanation:**

Centering problems choose representative interior points of convex sets. The Chebyshev center maximizes the radius of an inscribed ball, while analytic and ellipsoid centers use barrier or volume criteria. Feasibility keeps the centered object inside the set; optimality makes it as large or central as possible.

The variables are the candidate center $x_c$ and radius $r$. Maximizing $r$ means choosing the largest ball, centered at $x_c$, that remains inside the feasible set.

**Properties:**
- The Chebyshev radius is the distance to the nearest boundary.
- For an interval, the center is the midpoint.
- Higher-dimensional versions use convex inequalities.

**Intuition:**

The figure above shows center geometry; the calculation finds the largest interval ball.

<img src="../../../Figures/03020806_bv_fig_8_5_chebyshev_center.jpeg" alt="Boyd and Vandenberghe Figure 8.5: Chebyshev center of a polyhedron" style="width: 60%; height: auto;">


**Numerical Example:**

The figure above shows center geometry. For the interval $[0,4]$, choose center
$$
c=\frac{0+4}{2}=2.
$$

The distance to the left boundary is
$$
2-0=2,
$$
and the distance to the right boundary is
$$
4-2=2.
$$

Therefore the largest inscribed interval radius is $r=2$.


```python
import sympy as sp

left_boundary = sp.Integer(0)
right_boundary = sp.Integer(4)
center = (left_boundary + right_boundary) / 2
radius = min(center - left_boundary, right_boundary - center)

center, radius
```
