### 3.1.3. Necessary Optimality Conditions


$$
\nabla f(x^*) = 0,
\qquad
\nabla^2 f(x^*) \succeq 0.
$$


**Explanation:**

Necessary optimality conditions describe what must be true at a smooth local minimum. If $f$ is differentiable, then moving infinitesimally in any direction cannot produce a first-order decrease, so the gradient must vanish. If $f$ is twice differentiable, the second-order curvature cannot be negative in any direction, which is expressed by [positive semidefiniteness](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/18_matrix_definiteness.ipynb) of the Hessian in the sense of [matrix definiteness](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/18_matrix_definiteness.ipynb). These conditions are powerful filters, but they are not by themselves proof that a point is a minimum.

Here $\nabla f(x^*)$ is the gradient at the candidate point and $\nabla^2 f(x^*)$ is the Hessian matrix. The first condition says the first-order change is zero; the second says the Hessian is positive semidefinite in the sense of [matrix definiteness](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/18_matrix_definiteness.ipynb).

The distinction between necessary and sufficient conditions is one of the first conceptual hurdles in optimization. A stationary point can be a local minimum, a local maximum, or a saddle-like point, and a zero second derivative can leave the classification unresolved. These examples keep derivative tests from looking like automatic certificates.

**Intuition:**

The first figure shows that zero slope and nonnegative curvature can fail to identify a minimum when the function is not convex or when higher-order behavior matters. The second figure gives a more subtle warning: being minimal along every line through a point does not necessarily imply local minimality in the full space. The code cell reinforces the same lesson by comparing simple scalar functions that share stationary behavior but have different local meanings.

**Numerical Example:**

The figure warns that derivative tests are only filters. At $x^*=0$, compare three scalar costs:
$$
f_1(x)=x^2,\qquad f_2(x)=-x^2,\qquad f_3(x)=x^3.
$$

Their first and second derivatives at zero are
$$
\begin{array}{c|cc}
\text{cost} & f'(0) & f''(0) \\
\hline
x^2 & 0 & 2 \\
-x^2 & 0 & -2 \\
x^3 & 0 & 0
\end{array}
$$

For $f_1$, $f_1(-0.1)=0.01>0=f_1(0)$ and $f_1(0.1)=0.01>0$, so zero is a local minimum. For $f_2$, both nearby values are $-0.01<0$, so zero is a local maximum. For $f_3$, $f_3(-0.1)=-0.001<0<f_3(0.1)$, so the stationary point is neither a local minimum nor a local maximum.


```python
import sympy as sp

variable = sp.symbols("x")
objectives = {
    "minimum": variable**2,
    "maximum": -variable**2,
    "flat stationary point": variable**3,
}
second_order_tests = {
    name: (sp.diff(objective, variable).subs(variable, 0), sp.diff(objective, variable, 2).subs(variable, 0))
    for name, objective in objectives.items()
}

second_order_tests
```
