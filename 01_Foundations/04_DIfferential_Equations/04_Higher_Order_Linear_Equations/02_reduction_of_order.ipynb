### 1.4.4.2. Reduction of Order

$$
y = u(x)\,y_1(x) \;\Longrightarrow\;
y_2 = y_1(x)\int \frac{e^{-\int P(x)\,dx}}{y_1^2(x)}\,dx,
\qquad y'' + P(x)y' + Q(x)y = 0 .
$$

**Explanation:**

If one solution $y_1$ of a second-order linear homogeneous equation is known, **reduction of order** finds a second, independent solution $y_2$ by seeking $y = u(x)y_1(x)$. Substituting collapses the equation to a *first-order* equation in $w = u'$, which is [linear](../02_First_Order_Equations/04_linear_first_order_equations.ipynb) and separable; integrating twice gives $y_2$, summarized by the boxed formula. The pair $\{y_1, y_2\}$ is then a fundamental set, so $y = c_1 y_1 + c_2 y_2$. This idea also produces the repeated-root and Cauchy–Euler second solutions met later.

**Properties:**
- The substitution $w = u'$ turns a second-order equation into a solvable first-order one.
- $y_2/y_1 = \int e^{-\int P}/y_1^2\,dx$ is non-constant, guaranteeing independence.

**Numerical Example:**

Given that $y_1 = e^x$ solves $y'' - y = 0$, find a second solution.

**Set** $y = u(x)e^x$. By the product rule,

$$
y' = ue^x + e^x u', \qquad y'' = ue^x + 2e^x u' + e^x u''.
$$

**Substitute** into $y'' - y = 0$:

$$
ue^x + 2e^x u' + e^x u'' - ue^x = e^x(u'' + 2u') = 0 \;\Rightarrow\; u'' + 2u' = 0 .
$$

**Reduce order** with $w = u'$: the equation $w' + 2w = 0$ is linear/separable with $w = c_1 e^{-2x}$, so $u' = c_1 e^{-2x}$ and

$$
u = -\tfrac{c_1}{2}e^{-2x} + c_2 .
$$

**Back-substitute:** $y = u e^x = -\tfrac{c_1}{2}e^{-x} + c_2 e^x$. Choosing $c_1 = -2$, $c_2 = 0$ gives the second solution $y_2 = e^{-x}$. Since $W(e^x, e^{-x}) = -2 \neq 0$, the general solution is

$$
y = c_1 e^{x} + c_2 e^{-x}.
$$

```python
import sympy as sp

x = sp.symbols("x")
u = sp.Function("u")

substituted = (u(x) * sp.exp(x)).diff(x, 2) - u(x) * sp.exp(x)
reduced_equation = sp.simplify(substituted / sp.exp(x))

y_1 = sp.exp(x)
second_solution = y_1 * sp.integrate(sp.exp(-sp.integrate(0, x)) / y_1**2, x)
wronskian = sp.Matrix([[sp.exp(x), sp.exp(-x)], [sp.exp(x), -sp.exp(-x)]]).det()

print("after y = u e^x, equation becomes u'' + 2u' = 0 ->", reduced_equation, "= 0")
print("formula (5) gives y_2 =", second_solution)
print("W(e^x, e^-x) =", wronskian, " -> independent, general y = c1 e^x + c2 e^-x")
```
