### 3.9.5. Strong Convexity and Sublevel Geometry

$$mI\preceq\nabla^2f(x)\preceq MI$$


**Explanation:**

Strong convexity gives a lower curvature bound, while smoothness gives an upper curvature bound. Together they control sublevel-set geometry and algorithmic conditioning. Feasibility is unconstrained in this setting; optimality is governed by curvature around the minimizer.

The Hessian $\nabla^2 f(x)$ is bounded between two scalar multiples of the identity matrix $I$. In [matrix definiteness](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/18_matrix_definiteness.ipynb) terms, $mI\preceq\nabla^2 f(x)$ gives a lower curvature bound and $\nabla^2 f(x)\preceq MI$ gives an upper curvature bound.

**Properties:**
- The constants m and M bound Hessian [eigenvalues](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/08_eigenvalue.ipynb).
- The [condition number](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/31_condition_number.ipynb) is M/m.
- Large condition number means elongated sublevel sets.

**Intuition:**

The figure above shows conditioning geometry; the calculation reads m, M, and the condition number from eigenvalues.

<img src="../../Figures/030505_bv_fig_9_2_quadratic_condition_number.jpeg" alt="Boyd and Vandenberghe Figure 9.2: contours of a quadratic with condition number 10" style="width: 60%; height: auto;">


**Numerical Example:**

The figure above shows how conditioning changes sublevel geometry. Let
$$
H=\begin{bmatrix}2&1\\1&2\end{bmatrix}.
$$

The eigenvalues are $1$ and $3$, so
$$
m=1,\qquad M=3.
$$

The condition number is
$$
\kappa=\frac{M}{m}=3.
$$

Thus the quadratic has curvature between $1$ and $3$ in all directions.


```python
import sympy as sp

hessian = sp.Matrix([[2, 1], [1, 2]])
eigenvalues = hessian.eigenvals()
condition_number = sp.Integer(3)

print("eigenvalues =", eigenvalues)
print("condition_number =", condition_number)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

hessian = ca.DM([[2, 1], [1, 2]])
eigenvalues = ca.eig_symbolic(ca.SX(hessian))
condition_number = float(ca.mmax(eigenvalues) / ca.mmin(eigenvalues))

print("eigenvalues =", eigenvalues)
print("condition_number =", condition_number)
```
