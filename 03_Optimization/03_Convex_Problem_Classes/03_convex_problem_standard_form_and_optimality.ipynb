### 3.3.3. Convex Problem Standard Form and Optimality


$$\nabla f_0(x^*)^T(y-x^*)\geq0\quad\forall y\ \text{feasible}$$


**Explanation:**

In a convex problem, standard form makes the global optimality certificate explicit. A differentiable objective is optimal over the feasible set when its first-order inequality rules out every feasible descent direction.

The vector $x^*$ is the candidate optimum, and $y$ ranges over feasible points. The [inner product](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/02_inner_product_space.ipynb) $\nabla f_0(x^*)^T(y-x^*)$ measures the directional derivative from $x^*$ toward $y$ and must be nonnegative in every feasible direction.

**Properties:**
- The gradient defines the tangent approximation.
- Convexity keeps the true function above that approximation.
- At an optimum, the first-order inequality holds against every feasible point.

**Intuition:**

The computation compares a tangent lower bound with the true value.


**Numerical Example:**

Let $f(x)=x^2$, expand around $x=1$, and compare with $y=3$. Since $f'(x)=2x$,
$$
f'(1)=2.
$$

The tangent lower bound at $y=3$ is
$$
f(1)+f'(1)(3-1)=1^2+2\cdot2=5.
$$

The true value is
$$
f(3)=3^2=9.
$$

Because $9\geq5$, the tangent line gives a valid global lower bound.


```python
import sympy as sp

point_x = sp.Integer(1)
point_y = sp.Integer(3)
lower_bound = point_x ** 2 + 2 * point_x * (point_y - point_x)
true_value = point_y ** 2

print("lower_bound =", lower_bound)
print("true_value =", true_value)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

variable = ca.SX.sym("x")
objective = variable ** 2
gradient = ca.gradient(objective, variable)
point_x = 1
point_y = 3
function_at_x = ca.evalf(ca.substitute(objective, variable, point_x))
gradient_at_x = ca.evalf(ca.substitute(gradient, variable, point_x))
lower_bound = function_at_x + gradient_at_x * (point_y - point_x)
true_value = ca.evalf(ca.substitute(objective, variable, point_y))

print("lower_bound =", lower_bound)
print("true_value =", true_value)
```
