### 1.1.5.4. Distance Between Two Points


$$
d(\mathbf{p}, \mathbf{q}) = \lVert \mathbf{q} - \mathbf{p} \rVert
$$


**Explanation:**

The distance between two points is the Euclidean length of the displacement vector that connects them. It is the simplest distance formula in the module and follows directly from the [vector norm](../02_Vector/06_vector_norm.ipynb).

**Example:**

If

$$
\mathbf{p} =
\begin{bmatrix}
1 \\
2 \\
3
\end{bmatrix},
\qquad
\mathbf{q} =
\begin{bmatrix}
4 \\
6 \\
3
\end{bmatrix},
$$

then

$$
d(\mathbf{p}, \mathbf{q}) =
\left\lVert
\begin{bmatrix}
3 \\
4 \\
0
\end{bmatrix}
\right\rVert
= 5.
$$


```python
import sympy as sp

point_p = sp.Matrix([1, 2, 3])
point_q = sp.Matrix([4, 6, 3])
distance_value = (point_q - point_p).norm()

print("point_p =", list(point_p))
print("point_q =", list(point_q))
print("distance_value =", sp.simplify(distance_value))
```
