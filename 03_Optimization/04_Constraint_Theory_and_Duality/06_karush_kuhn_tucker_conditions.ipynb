### 3.4.6. Karush-Kuhn-Tucker Conditions

$$
\begin{aligned}
\nabla_x L(x^*,\lambda^*,\mu^*) &= 0,\\
h(x^*) &= 0,\\
g_j(x^*) &\le 0,\\
\mu_j^* &\ge 0,\\
\mu_j^*g_j(x^*) &= 0.
\end{aligned}
$$

**Explanation:**

For inequality constraints, active constraints behave like equalities at the solution, while inactive constraints have zero multipliers. The KKT conditions combine stationarity, primal feasibility, nonnegative inequality multipliers, and complementary slackness. The nonnegativity of $\mu_j^*$ reflects that relaxing an active inequality can only decrease the optimal cost to first order.

**Intuition:**

<img src="../../Figures/030406_berts_fig_3_3_1_inequality_multiplier_nonnegative.jpeg" alt="Bertsekas Figure 3.3.1: nonnegative multiplier for an active inequality constraint" style="width: 60%; height: auto;">

The figure shows why the active inequality multiplier is nonnegative: the objective gradient points inward, while the active constraint gradient points outward.

**Numerical Example:**

Consider

$$
\min \frac{1}{2}(x_1^2+x_2^2+x_3^2)
\quad \text{subject to} \quad g(x)=x_1+x_2+x_3+3\le 0.
$$

The Lagrangian is

$$
L(x,\mu)=\frac{1}{2}(x_1^2+x_2^2+x_3^2)+\mu(x_1+x_2+x_3+3).
$$

Stationarity gives

$$
x_1+\mu=0,
\qquad x_2+\mu=0,
\qquad x_3+\mu=0,
$$

so

$$
x_1=x_2=x_3=-\mu.
$$

The unconstrained minimizer $x=(0,0,0)$ is infeasible because $g(0)=3>0$, so the inequality must be active:

$$
x_1+x_2+x_3+3=0.
$$

Substitute $x_i=-\mu$:

$$
-3\mu+3=0
\quad \Rightarrow \quad
\mu=1.
$$

Therefore

$$
x^*=(-1,-1,-1), \qquad \mu^*=1.
$$

The constraint is active, the multiplier is nonnegative, and complementary slackness holds because $\mu^*g(x^*)=1\cdot 0=0$.

```python
import sympy as sp

x_1, x_2, x_3, mu = sp.symbols('x_1 x_2 x_3 mu')

cost = sp.Rational(1, 2) * (x_1**2 + x_2**2 + x_3**2)
constraint = x_1 + x_2 + x_3 + 3
lagrangian = cost + mu * constraint
stationarity_equations = [
    sp.diff(lagrangian, x_1),
    sp.diff(lagrangian, x_2),
    sp.diff(lagrangian, x_3),
    constraint,
]
solution = sp.solve(stationarity_equations, [x_1, x_2, x_3, mu], dict=True)[0]

result = {
    'solution': solution,
    'constraint_value': constraint.subs(solution),
    'complementary_slackness': (mu * constraint).subs(solution),
    'cost_at_solution': cost.subs(solution),
}

for label, value in result.items():
    print(f"{label} =", value)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

opti = ca.Opti()
decision_variable = opti.variable(3)
opti.minimize(0.5 * ca.dot(decision_variable, decision_variable))
opti.subject_to(decision_variable[0] + decision_variable[1] + decision_variable[2] == -3)
opti.solver("ipopt", {"print_time": False}, {"print_level": 0, "sb": "yes"})
solution = opti.solve()

print("minimizer =", ca.DM(solution.value(decision_variable)))
```

**Convex problems — KKT as a sufficient condition:**

In general the KKT system is only *necessary* for optimality. When the problem is convex — $f_0$ and the $f_i$ convex, the $h_i$ affine — the same conditions become *sufficient*: any point $(x^*,\lambda^*,\mu^*)$ satisfying stationarity, primal and dual feasibility, and complementary slackness is a global minimizer,

$$
\nabla f_0(x^*)+\sum_i \mu_i^*\,\nabla f_i(x^*)+\sum_i \lambda_i^*\,\nabla h_i(x^*)=0.
$$

This is why KKT points certify global optimality for the convex [problem classes](../03_Convex_Problem_Classes/03_convex_problem_standard_form_and_optimality.ipynb) and underpins the duality results that follow.
