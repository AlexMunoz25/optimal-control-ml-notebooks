### 1.1.5.13. Vector Space


$$
(V, \mathbb{R}, +, \cdot)
$$


**Explanation:**

A vector space is a set closed under vector addition and scalar multiplication, together with the usual linear-algebra axioms. This structure explains why lines, planes, spans, and matrix subspaces all fit the same pattern. It is the parent notion for [span](./14_span.ipynb) and [subspace](./15_vector_subspace.ipynb).

**Example:**

The set $\mathbb{R}^3$ is a vector space because adding two three-dimensional vectors stays in $\mathbb{R}^3$, and multiplying a three-dimensional vector by a real scalar also stays in $\mathbb{R}^3$.


```python
import sympy as sp

vector_u = sp.Matrix([1, 2, 3])
vector_v = sp.Matrix([4, -1, 2])
scalar_value = 3

print("vector_addition =", vector_u + vector_v)
print("scalar_multiplication =", scalar_value * vector_u)
```
