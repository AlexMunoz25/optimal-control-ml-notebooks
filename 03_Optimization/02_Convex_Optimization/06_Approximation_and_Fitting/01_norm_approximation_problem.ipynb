### 3.2.6.1. Norm Approximation Problem


$$\min_x\|Ax-b\|$$


**Explanation:**

[Norm](../../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb) approximation chooses decision variables so the residual vector is small in a specified norm. The selected norm determines the geometry of the error measure and keeps the objective convex when composed with an affine residual.

The [matrix](../../../01_Foundations/01_Linear_Algebra/03_Matrix/01_matrix_definition.ipynb) $A$ maps the decision vector $x$ to the model output $Ax$. The vector $b$ is the target data, and the objective minimizes the [norm](../../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb) of the residual $Ax-b$.

**Properties:**
- Residuals measure mismatch between a model and data.
- Squaring residuals gives a smooth convex objective.
- The derivative or normal equations give the optimality condition.

**Intuition:**

The computation expands the residual objective so the convex quadratic structure is visible before solving.


**Numerical Example:**

Let $r_1(x)=x-1$ and $r_2(x)=2x-3$. The least-squares objective is
$$
f(x)=(x-1)^2+(2x-3)^2.
$$

Expand it:
$$
f(x)=x^2-2x+1+4x^2-12x+9=5x^2-14x+10.
$$

Differentiate and set the derivative to zero:
$$
f'(x)=10x-14=0.
$$

Thus
$$
x^\star=\frac{14}{10}=\frac75.
$$

The minimum value is
$$
f(7/5)=(2/5)^2+(-1/5)^2=\frac15.
$$


```python
import sympy as sp

variable = sp.symbols('x')
objective = (variable - 1) ** 2 + (2 * variable - 3) ** 2
optimizer = sp.solve(sp.diff(objective, variable), variable)[0]
minimum_value = sp.simplify(objective.subs(variable, optimizer))

print("optimizer =", optimizer)
print("minimum_value =", minimum_value)
```
