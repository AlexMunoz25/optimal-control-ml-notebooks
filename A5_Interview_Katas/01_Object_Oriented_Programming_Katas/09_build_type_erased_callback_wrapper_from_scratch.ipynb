### A5.1.9. Build Type Erased Callback Wrapper from Scratch

**Problem:**

Build a wrapper class that can store any callable (function, lambda, bound method, or callable object) and invoke it later with a uniform interface.

**Explanation:**

Type erasure means hiding the concrete type of a callable behind a uniform wrapper. The wrapper stores any callable and calls it through a single `__call__` method. The caller does not need to know whether it holds a function, a lambda, or a method.

How to build it:

1. Store the callable in `__init__`.
2. Implement `__call__(self, *args, **kwargs)` to forward all arguments to the stored callable.
3. The wrapper behaves like a regular function from the outside — anything that holds the wrapper can call it without knowing the original type.

**Example:**

Wrap a regular function, a lambda, and a bound method in the same `Callback` class. All three are invoked identically through `callback(args)`.

```python
class Callback:
    def __init__(self, callable_obj):
        self.callable_obj = callable_obj

    def __call__(self, *args, **kwargs):
        return self.callable_obj(*args, **kwargs)


def greet(name):
    return f"Hello, {name}!"


class Printer:
    def print_message(self, message):
        return f"Printing: {message}"


callback_function = Callback(greet)
callback_lambda = Callback(lambda x: x * 2)
callback_method = Callback(Printer().print_message)

print(callback_function("Alice"))
print(callback_lambda(21))
print(callback_method("test"))

callbacks = [callback_function, callback_lambda, callback_method]
for callback in callbacks:
    print(type(callback).__name__, "→", callback("demo"))
```
