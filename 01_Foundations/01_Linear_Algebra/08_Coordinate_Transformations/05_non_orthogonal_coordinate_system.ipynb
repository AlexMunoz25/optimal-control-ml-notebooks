### 1.1.8.5. Non-Orthogonal Coordinate System


$$
\mathbf{A} = A^1 \mathbf{e}_1 + A^2 \mathbf{e}_2,
\qquad
\mathbf{A} \neq A_1 \mathbf{e}_1 + A_2 \mathbf{e}_2
$$


**Explanation:**

When the basis directions are not perpendicular, projecting parallel to one axis is no longer the same operation as projecting perpendicular to the other axis. The parallel projections give coefficients that rebuild the vector with the original basis vectors, but the perpendicular shadows do not. This is why skew coordinate systems require two distinct component sets: contravariant coefficients and covariant projections.

**Example:**

With
$$
\mathbf{e}_1 =
\begin{bmatrix}
1 \\
3
\end{bmatrix},
\qquad
\mathbf{e}_2 =
\begin{bmatrix}
4 \\
0
\end{bmatrix},
\qquad
\mathbf{A} =
\begin{bmatrix}
7 \\
2
\end{bmatrix},
$$
solving $\mathbf{A} = A^1 \mathbf{e}_1 + A^2 \mathbf{e}_2$ gives the parallel-projection coefficients
$$
A^1 = \frac{2}{3},
\qquad
A^2 = \frac{19}{12}.
$$
The perpendicular projections lead to a different pair of numbers, so the two notions must be kept separate.


```python
import sympy as sp

basis_vector_1 = sp.Matrix([1, 3])
basis_vector_2 = sp.Matrix([4, 0])
vector_a = sp.Matrix([7, 2])
basis_matrix = sp.Matrix.hstack(basis_vector_1, basis_vector_2)
parallel_projection_coefficients = basis_matrix.LUsolve(vector_a)
perpendicular_projection_numbers = [vector_a.dot(basis_vector) for basis_vector in [basis_vector_1, basis_vector_2]]
incorrect_reconstruction = (
    perpendicular_projection_numbers[0] * basis_vector_1
    + perpendicular_projection_numbers[1] * basis_vector_2
)

print("parallel_projection_coefficients =", parallel_projection_coefficients)
print("perpendicular_projection_numbers =", perpendicular_projection_numbers)
print("incorrect_reconstruction_from_original_basis =", incorrect_reconstruction)
```
