### 1.3.1.6. Intermediate Value Theorem

$$
f \text{ continuous on } [a,b], \;\; z \text{ between } f(a) \text{ and } f(b)
\;\;\Longrightarrow\;\; \exists\, c \in [a,b] : f(c) = z .
$$

**Explanation:**

A function continuous on a closed interval $[a,b]$ takes on every value between $f(a)$ and $f(b)$ at some point of the interval. The case used most in computation is $z = 0$: if $f(a)$ and $f(b)$ have opposite signs, then $f$ has a root somewhere in $(a,b)$. This existence guarantee is the theoretical basis of bracketing root-finders such as bisection, which repeatedly halve an interval known to contain a sign change — the robust counterpart to the locally-fast [Newton's method](../02_Single_Variable_Differentiation/11_newtons_method_for_root_finding.ipynb).

**Intuition:**

The continuous graph cannot pass from below the line $y = z$ to above it without crossing it.

**Numerical Example:**

Let $f(x) = x - \cos x$, which is continuous everywhere. At the endpoints of $[0, \pi/2]$,

$$
f(0) = 0 - \cos 0 = -1 < 0, \qquad f\!\left(\tfrac{\pi}{2}\right) = \tfrac{\pi}{2} - \cos\tfrac{\pi}{2} = \tfrac{\pi}{2} > 0 .
$$

The signs differ, so by the theorem there is a root $c \in (0, \pi/2)$. One bisection step evaluates the midpoint $m = \pi/4 \approx 0.785$:

$$
f\!\left(\tfrac{\pi}{4}\right) = \tfrac{\pi}{4} - \tfrac{\sqrt2}{2} \approx 0.785 - 0.707 = 0.078 > 0,
$$

so the sign change is now on $(0, \pi/4)$ and the root is bracketed more tightly. Iterating converges to $c \approx 0.739$.

```python
import sympy as sp

x = sp.symbols("x")
f = x - sp.cos(x)

left_end, right_end = sp.Integer(0), sp.pi / 2
left_value = f.subs(x, left_end)
right_value = sp.N(f.subs(x, right_end))
sign_change = sp.sign(left_value) != sp.sign(right_value)

midpoint = sp.pi / 4
midpoint_value = sp.N(f.subs(x, midpoint))
root = sp.nsolve(f, x, 0.75)

print("f(0)     =", left_value)
print("f(pi/2)  =", right_value, " -> opposite signs:", sign_change)
print("f(pi/4)  =", midpoint_value, " -> root now bracketed in (0, pi/4)")
print("root c   =", root)
```

**Equivalent CasADi implementation:**

The same sign-bracketed root is the natural job of a numerical root-finder. CasADi builds a Newton root-finder from the function graph and returns the same $c$.

```python
import casadi as ca

x = ca.SX.sym("x")
residual = x - ca.cos(x)
root_finder = ca.rootfinder("root_finder", "newton", {"x": x, "g": residual})

solution = root_finder(x0=0.75)
print("CasADi root c =", float(solution["x"]))
```
