### 3.2.10.8. Structured Examples: Network Flow and Optimal Control


$$\min_x\sum_i\phi_i(x_i)\quad\text{subject to}\quad Ax=b$$


**Explanation:**

Structured equality constraints appear in network flow, trajectory, and optimal-control examples, where conservation or dynamics creates sparse linear equations. Feasibility is the equality constraint or its linearized version; optimality is stationarity together with that feasibility condition.

The objective is separable because it sums scalar terms $\phi_i(x_i)$. The equality constraint $Ax=b$ couples those variables, so the optimization structure comes from a separable cost plus a sparse linear constraint.

**Properties:**
- Equality constraints introduce unrestricted multipliers.
- KKT systems couple stationarity and feasibility.
- Eliminating constraints can reduce the number of variables.

**Intuition:**

The numerical example keeps the equality constraint visible in every algebraic step.


**Numerical Example:**

A two-arc flow split with total demand $2$ can be written
$$
f_1+f_2=2.
$$

Minimizing
$$
\frac12(f_1^2+f_2^2)
$$
under this equality gives stationarity
$$
f_1+\nu=0,
\qquad
f_2+\nu=0.
$$

Thus $f_1=f_2=-\nu$. The demand equation gives
$$
-2\nu=2\quad\Longrightarrow\quad\nu=-1,
$$
so $f_1=f_2=1$.


```python
import sympy as sp

flow_1, flow_2, multiplier = sp.symbols('f_1 f_2 nu')
solution = sp.solve([sp.Eq(flow_1 + multiplier, 0), sp.Eq(flow_2 + multiplier, 0), sp.Eq(flow_1 + flow_2, 2)], [flow_1, flow_2, multiplier])

solution
```
