### 3.3.18. Operations Preserving Convexity of Functions

$$f=\alpha_1f_1+\cdots+\alpha_kf_k,\qquad\alpha_i\geq0$$


**Explanation:**

Convexity-preserving operations make it possible to build complicated [convex functions](01_convex_function_definition_and_examples.ipynb) from simpler ones. Nonnegative sums, affine composition, pointwise maxima, and related rules avoid rechecking the definition from scratch.

The display forms a new function by adding functions $f_i$ with nonnegative weights $\alpha_i$. Nonnegative weighting is the part that preserves convexity when the functions being added are convex.

**Properties:**
- The input average is compared with the average of function values.
- Quadratics with nonnegative curvature are canonical convex examples.
- Jensen inequalities extend the same idea to many points.

**Intuition:**

The computation evaluates both sides of the convexity inequality.


**Numerical Example:**

Let $f(x)=x^2$, $x=0$, $y=2$, and $\theta=1/4$. The averaged input is
$$
\theta x+(1-\theta)y=\frac14\cdot0+\frac34\cdot2=\frac32.
$$

The left side is
$$
f(3/2)=(3/2)^2=9/4.
$$

The right side is
$$
\frac14 f(0)+\frac34 f(2)=\frac14\cdot0+\frac34\cdot4=3.
$$

Since $9/4\leq3$, the convexity inequality holds.


```python
import sympy as sp

theta = sp.Rational(1, 4)
point_x = sp.Integer(0)
point_y = sp.Integer(2)
midpoint = theta * point_x + (1 - theta) * point_y
left_side = midpoint ** 2
right_side = theta * point_x ** 2 + (1 - theta) * point_y ** 2

print("left_side =", left_side)
print("right_side =", right_side)
print("left_side <= right_side =", left_side <= right_side)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

theta = 0.25
point_x = ca.DM(0)
point_y = ca.DM(2)
midpoint = theta * point_x + (1 - theta) * point_y
left_side = midpoint ** 2
right_side = theta * point_x ** 2 + (1 - theta) * point_y ** 2

print("left_side =", left_side)
print("right_side =", right_side)
print("left_side <= right_side =", bool(left_side <= right_side))
```
