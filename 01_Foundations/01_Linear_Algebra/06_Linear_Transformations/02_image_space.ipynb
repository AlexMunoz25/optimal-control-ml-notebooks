### 1.1.6.2. Image Space


$$
\operatorname{Im}(T) =
\left\{
\mathbf{w} \in W \mid \mathbf{w} = T(\mathbf{v}) \text{ for some } \mathbf{v} \in V
\right\}
$$


**Explanation:**

The image space is the set of all vectors a linear transformation can output. It is the transformation-side counterpart of the [column space](../05_Geometrical_Aspects_of_Linear_Algebra/16_matrix_fundamental_spaces.ipynb), so it captures exactly which outputs are reachable. Whether the image fills the whole codomain is the question addressed by [injective, surjective, and bijective transformations](./04_injective_surjective_bijective.ipynb).

**Example:**

For

$$
T(x, y) =
\begin{bmatrix}
x \\
y \\
x + y
\end{bmatrix},
$$

the image space is

$$
\operatorname{Im}(T) =
\operatorname{span}
\left\{
\begin{bmatrix}
1 \\
0 \\
1
\end{bmatrix},
\begin{bmatrix}
0 \\
1 \\
1
\end{bmatrix}
\right\}.
$$


```python
import sympy as sp

transformation_matrix = sp.Matrix([[1, 0], [0, 1], [1, 1]])
image_basis = transformation_matrix.columnspace()

print("image_basis =", image_basis)
```
