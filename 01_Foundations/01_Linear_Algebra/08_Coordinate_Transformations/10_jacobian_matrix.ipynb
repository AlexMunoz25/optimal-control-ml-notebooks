### 1.1.8.10. Jacobian Matrix


$$
J^{i'}_{\;j} = \frac{\partial x^{i'}}{\partial x^j},
\qquad
dx^{i'} = \frac{\partial x^{i'}}{\partial x^j} \, dx^j,
\qquad
A^{i'} = \frac{\partial x^{i'}}{\partial x^j} A^j,
\qquad
A_{i'} = \frac{\partial x^j}{\partial x^{i'}} A_j
$$


**Explanation:**

For a nonlinear coordinate change, the Jacobian is the local linear map between differential coordinate increments. Its columns are the tangent basis vectors of the original coordinates written in the new coordinates, so the Jacobian gives the transformation law for contravariant quantities. The inverse Jacobian gives the matching law for covariant quantities.

**Example:**

For polar-to-Cartesian coordinates,
$$
x = r \cos \theta,
\qquad
y = r \sin \theta,
\qquad
J(r, \theta) =
\begin{bmatrix}
\cos \theta & -r \sin \theta \\
\sin \theta & r \cos \theta
\end{bmatrix}.
$$
The first column is the radial tangent direction, and the second column is the tangential direction scaled by $r$.


```python
import sympy as sp

radial_coordinate, angular_coordinate = sp.symbols("radial_coordinate angular_coordinate", positive=True)
cartesian_x = radial_coordinate * sp.cos(angular_coordinate)
cartesian_y = radial_coordinate * sp.sin(angular_coordinate)
jacobian_matrix = sp.Matrix(
    [
        [sp.diff(cartesian_x, radial_coordinate), sp.diff(cartesian_x, angular_coordinate)],
        [sp.diff(cartesian_y, radial_coordinate), sp.diff(cartesian_y, angular_coordinate)],
    ]
)
point_substitution = {radial_coordinate: 2, angular_coordinate: sp.pi / 6}

print("jacobian_matrix =")
sp.pprint(jacobian_matrix)
print("jacobian_matrix_at_r_2_theta_pi_over_6 =")
sp.pprint(sp.simplify(jacobian_matrix.subs(point_substitution)))
```
