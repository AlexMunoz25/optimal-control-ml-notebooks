### 3.6.16. SOCP and SDP Barrier Examples


$$-\log(t^2-x^Tx),\qquad -\log\det X$$


**Explanation:**

SOCP and SDP barriers are concrete [cone barrier](../02_Convex_Analysis/03_cones_and_conic_combinations.ipynb) examples. The second-order cone barrier uses the positive slack t^2-x^Tx, while the semidefinite cone barrier uses [determinant](../../01_Foundations/01_Linear_Algebra/03_Matrix/10_matrix_determinant.ipynb) positivity.

The first barrier is for a second-order cone slack with scalar $t$ and vector $x$. The second barrier is for a [positive definite matrix](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/17_positive_definite_matrix.ipynb) $X$, where $\det X$ measures distance to the semidefinite boundary in determinant form.

**Properties:**
- SOCP strict feasibility requires t to dominate the [norm](../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb).
- SDP strict feasibility requires positive definiteness.
- Both barriers diverge near the cone boundary.

**Intuition:**

The example evaluates one SOCP slack and one SDP log-det barrier.


**Numerical Example:**

For an SOCP point, let
$$u=\begin{bmatrix}3\\4\end{bmatrix},\qquad t=6.
$$

The strict cone slack is
$$
t^2-u^Tu=6^2-(3^2+4^2)=36-25=11.
$$

The SOCP barrier contribution is $-\log(11)$. For
$$
X=\begin{bmatrix}2&1\\1&2\end{bmatrix},
$$
$\det X=3$, so the SDP barrier contribution is $-\log3$.


```python
import sympy as sp

vector = sp.Matrix([3, 4])
height = sp.Integer(6)
socp_slack = height ** 2 - (vector.T * vector)[0]
socp_barrier = -sp.log(socp_slack)
matrix = sp.Matrix([[2, 1], [1, 2]])
sdp_barrier = -sp.log(matrix.det())

print("socp_slack =", socp_slack)
print("socp_barrier =", socp_barrier)
print("sdp_barrier =", sdp_barrier)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

vector = ca.DM([3, 4])
height = 6
socp_slack = height ** 2 - ca.dot(vector, vector)
socp_barrier = -ca.log(socp_slack)
matrix = ca.DM([[2, 1], [1, 2]])
sdp_barrier = -ca.log(ca.det(matrix))

print("socp_slack =", socp_slack)
print("socp_barrier =", socp_barrier)
print("sdp_barrier =", sdp_barrier)
```
