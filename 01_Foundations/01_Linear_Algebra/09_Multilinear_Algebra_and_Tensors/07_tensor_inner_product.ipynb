### 1.1.9.7. Tensor Inner Product

$$
A^i B_i = \text{scalar}
$$

$$
C^i{}_k = A^{ij} B_{jk}
$$

$$
\text{inner product} = \text{outer product} + \text{contraction}
$$


**Explanation:**

- The inner product first forms an outer product and then contracts one matching index pair.
- The dot product and matrix multiplication are the familiar low-rank cases.
- Each contraction lowers the total rank by two.

**Connections:**

- This unifies the earlier dot-product and matrix-multiplication notebooks under one tensor rule.
- In TensorFlow, `tf.matmul` and many `tf.einsum` calls are tensor inner products.

**Example:**

- $A^{ij} B_{jk}$ contracts over $j$ and leaves indices $i$ and $k$, so the result is rank 2.
- $A^i B_i$ contracts the only index pair and leaves a scalar.


```python
import numpy as np

vector_a = np.array([1.0, 2.0, 3.0])
vector_b = np.array([4.0, 5.0, 6.0])
matrix_a = np.array([[1.0, 2.0], [3.0, 4.0]])
matrix_b = np.array([[5.0, 6.0], [7.0, 8.0]])

vector_inner_product = np.einsum('i,i->', vector_a, vector_b)
matrix_inner_product = np.einsum('ij,jk->ik', matrix_a, matrix_b)

print("vector_inner_product =", vector_inner_product)
print("matrix_inner_product =\n", matrix_inner_product)
print("matrix_a @ matrix_b =\n", matrix_a @ matrix_b)
```
