### 1.3.5.7. Jacobian

$$
\mathbf{F} : \mathbb{R}^{n} \to \mathbb{R}^{m}, \qquad
J(\mathbf{x}) = \frac{\partial \mathbf{F}}{\partial \mathbf{x}} =
\begin{bmatrix}
\dfrac{\partial F_1}{\partial x_1} & \cdots & \dfrac{\partial F_1}{\partial x_n} \\
\vdots & \ddots & \vdots \\
\dfrac{\partial F_m}{\partial x_1} & \cdots & \dfrac{\partial F_m}{\partial x_n}
\end{bmatrix}.
$$

**Explanation:**

The Jacobian is the derivative of a vector-valued map: row $i$ is the [gradient](./03_gradient.ipynb) of the $i$th component function, so the whole matrix collects every first-order partial derivative of $\mathbf{F}$. It is the best linear map approximating $\mathbf{F}$ near a point, generalizing the scalar derivative, and acting on a displacement by ordinary [matrix–vector multiplication](../../01_Linear_Algebra/03_Matrix/04_matrix_vector_multiplication.ipynb). Jacobians are the linearization at the heart of state-space linearization in control and the [Gauss–Newton](../../../03_Optimization/09_Algorithms/19_gauss_newton_method.ipynb) method for nonlinear least squares.

**Properties:**
- For $m = 1$ the Jacobian is the $1 \times n$ transpose of the gradient.
- First-order approximation: $\mathbf{F}(\mathbf{x}) \approx \mathbf{F}(\mathbf{x}_0) + J(\mathbf{x}_0)(\mathbf{x} - \mathbf{x}_0)$.

**Numerical Example:**

Let $\mathbf{F}(x, y) = \begin{bmatrix} x^2 + y \\ xy \end{bmatrix}$. Differentiating each component with respect to $x$ and $y$,

$$
J(x, y) = \begin{bmatrix} 2x & 1 \\ y & x \end{bmatrix}, \qquad
J(1, 2) = \begin{bmatrix} 2 & 1 \\ 2 & 1 \end{bmatrix}.
$$

At $(1,2)$ the determinant is $2\cdot 1 - 1\cdot 2 = 0$, so the linearization there is singular — $\mathbf{F}$ locally collapses a direction (it is not invertible at that point).

```python
import sympy as sp

x, y = sp.symbols("x y")
vector_map = sp.Matrix([x**2 + y, x * y])
variables = sp.Matrix([x, y])

jacobian = vector_map.jacobian(variables)
jacobian_at_point = jacobian.subs({x: 1, y: 2})
jacobian_determinant = jacobian_at_point.det()

print("J(x,y) ="); sp.pprint(jacobian)
print("J(1,2) ="); sp.pprint(jacobian_at_point)
print("det J(1,2) =", jacobian_determinant)
```

**Equivalent CasADi implementation:**

`ca.jacobian` returns the same matrix of partials of the vector map.

```python
import casadi as ca

variables = ca.SX.sym("x", 2)
vector_map = ca.vertcat(variables[0]**2 + variables[1], variables[0] * variables[1])
jacobian = ca.jacobian(vector_map, variables)
jacobian_function = ca.Function("jacobian_map", [variables], [jacobian])

print("J(1,2) =", jacobian_function([1, 2]))
```
