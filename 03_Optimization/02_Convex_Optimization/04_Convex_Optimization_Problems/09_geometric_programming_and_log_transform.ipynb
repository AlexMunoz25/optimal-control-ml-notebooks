### 3.2.4.9. Geometric Programming and Log Transform


$$f(x)=\sum_k c_kx_1^{a_{1k}}\cdots x_n^{a_{nk}},\qquad c_k>0$$


**Explanation:**

Geometric programs use positive variables, monomials, and posynomials. After the logarithmic change of variables, monomials become affine and posynomials become log-sum-exp expressions, which are convex.

The expression is a posynomial: a sum of monomial terms with positive coefficients $c_k$. Each exponent $a_{ik}$ controls how the positive variable $x_i$ enters the $k$th monomial.

**Properties:**
- Variables must be positive.
- Monomial products become affine in log variables.
- Posynomial inequalities become convex after the log transform.

**Intuition:**

The example evaluates a monomial and the equivalent log-domain expression.


**Numerical Example:**

Consider the monomial
$$
f(x_1,x_2)=4x_1x_2^2
$$
at $x_1=2$, $x_2=3$. In original variables,
$$
f(2,3)=4\cdot2\cdot3^2=72.
$$

Let $u_i=\log x_i$. Then
$$
\log f=\log4+u_1+2u_2.
$$

Substitute $u_1=\log2$ and $u_2=\log3$:
$$
\log f=\log4+\log2+2\log3=\log72.
$$

Exponentiating returns $72$.


```python
import sympy as sp

value_1 = sp.Integer(2)
value_2 = sp.Integer(3)
monomial_value = 4 * value_1 * value_2 ** 2
log_value = sp.log(4) + sp.log(value_1) + 2 * sp.log(value_2)
recovered_value = sp.simplify(sp.exp(log_value))

monomial_value, recovered_value
```
