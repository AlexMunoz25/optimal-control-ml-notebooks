### 3.4.7. Slack Variable Conversion and Inequality Sufficiency

$$
g_j(x)+z_j^2=0, \qquad z_j=\sqrt{-g_j(x)}
$$

$$
y^T\nabla_{xx}^2L(x^*,\lambda^*,\mu^*)y>0
$$

**Explanation:**

An inequality $g_j(x)\le 0$ can be converted into an equality by adding a squared slack variable, $g_j(x)+z_j^2=0$. This converts inequality-constrained conditions into equality-constrained ones and explains why inactive constraints receive zero multipliers. For second-order sufficiency in the inequality case, the strict curvature condition is imposed along directions tangent to the equality constraints and active inequalities, together with strict complementary slackness for active inequalities.

**Assumptions:**
- The functions are twice continuously differentiable for the second-order statement.
- Strict complementary slackness requires $\mu_j^*>0$ for active inequality constraints in this sufficient condition.

**Numerical Example:**

Use the one-dimensional inequality

$$
g(x)=x-1\le 0.
$$

At the feasible point $x^*=0$,

$$
g(x^*)=-1<0,
$$

so the constraint is inactive. The slack conversion gives

$$
g(x^*)+z^{*2}=0
\quad \Rightarrow \quad
-1+z^{*2}=0,
$$

hence

$$
z^*=1.
$$

The equality-converted stationarity condition with respect to $z$ contains

$$
2\mu^* z^*=0.
$$

Because $z^*=1$, this gives

$$
2\mu^*=0 \quad \Rightarrow \quad \mu^*=0.
$$

Thus the slack-variable conversion reproduces the KKT rule: inactive inequalities have zero multipliers.

```python
import sympy as sp

x, slack, mu = sp.symbols('x z mu')

constraint = x - 1
candidate = {x: 0}
slack_solution = sp.sqrt(-constraint.subs(candidate))
slack_stationarity = 2 * mu * slack
mu_solution = sp.solve(slack_stationarity.subs(slack, slack_solution), [mu], dict=True)[0]

result = {
    'constraint_at_candidate': constraint.subs(candidate),
    'slack_solution': slack_solution,
    'mu_solution': mu_solution,
    'converted_constraint_value': (constraint + slack**2).subs({x: 0, slack: slack_solution}),
}

for label, value in result.items():
    print(f"{label} =", value)
```
