### 3.2.7.6. Maximum Entropy and KL Divergence


$$H(p)=-\sum_i p_i\log p_i,
\qquad D(p\|q)=\sum_i p_i\log(p_i/q_i)$$


**Explanation:**

Entropy is a concave function on the probability simplex. Maximizing entropy under convex probability constraints is a standard convex-optimization form after switching to concave maximization. Feasibility is nonnegative probabilities summing to one; optimality favors spread-out distributions unless constraints force otherwise.

**Properties:**
- Probabilities must be nonnegative and sum to one.
- Entropy rewards spread-out mass.
- KL divergence and maximum entropy are central statistical convex models.

**Intuition:**

The example computes entropy for the two-outcome uniform distribution.

<img src="../../../Figures/03020706_bv_fig_7_2_maximum_entropy_distribution.jpeg" alt="Boyd and Vandenberghe Figure 7.2: maximum entropy distribution satisfying constraints" style="width: 60%; height: auto;">


**Numerical Example:**

The figure above gives the visual version of the same calculation.

Let
$$
p=\begin{bmatrix}1/2\\1/2\end{bmatrix}.
$$

The entropy is
$$
H(p)=-\frac12\log\frac12-\frac12\log\frac12.
$$

Combine the equal terms:
$$
H(p)=-\log\frac12.
$$

Since $-\log(1/2)=\log2$, the entropy is $\log2$.


```python
import sympy as sp

probabilities = [sp.Rational(1, 2), sp.Rational(1, 2)]
entropy = -sum(probability * sp.log(probability) for probability in probabilities)

sp.simplify(entropy)
```
