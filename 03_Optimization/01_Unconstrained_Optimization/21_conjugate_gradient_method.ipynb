### 3.1.21. Conjugate Gradient Method


$$
\begin{aligned}
g^k &= Qx^k-b, \\
\alpha^k &= \frac{g^{k\mathsf T}g^k}{d^{k\mathsf T}Qd^k}, \\
x^{k+1} &= x^k+\alpha^k d^k, \\
d^{k+1} &= -g^{k+1}+\beta^k d^k.
\end{aligned}
$$


**Explanation:**

The conjugate gradient method generates conjugate directions without first constructing a full conjugate basis. For a positive definite quadratic, it uses gradients and previous search directions to build the next direction, and in exact arithmetic it terminates in at most $n$ steps. Its major practical advantage is that it needs matrix-vector products rather than a factorization of $Q$, making it attractive for large problems.

Conceptually, conjugate gradient improves on steepest descent by remembering useful directional information. The residuals become orthogonal in the Euclidean [inner product](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/02_inner_product_space.ipynb), while the search directions become conjugate in the $Q$ geometry. This combination explains why the method can be much faster than steepest descent while using far less storage than Newton or quasi-Newton methods.

**Example:**

In two dimensions, exact conjugate gradient can solve a positive definite quadratic in two iterations. The code cell demonstrates this finite-termination property on a small quadratic system. The example is intentionally small so the reader can connect the algebraic updates with the geometric claim about conjugate directions.


```python
import numpy as np

Q = np.array([[4.0, 1.0], [1.0, 3.0]])
linear_term = np.array([1.0, 2.0])
current_point = np.array([0.0, 0.0])
gradient = Q @ current_point - linear_term
direction = -gradient

for iteration in range(2):
    stepsize = (gradient @ gradient) / (direction @ Q @ direction)
    current_point = current_point + stepsize * direction
    next_gradient = Q @ current_point - linear_term
    beta = (next_gradient @ next_gradient) / (gradient @ gradient)
    direction = -next_gradient + beta * direction
    gradient = next_gradient

current_point
```
