### 1.3.3.3. Fundamental Theorem of Calculus

$$
\textbf{Part 1:}\quad \frac{d}{dx}\int_{a}^{x} f(t)\,dt = f(x), \qquad
\textbf{Part 2:}\quad \int_{a}^{b} f(x)\,dx = F(b) - F(a),\ \ F' = f .
$$

**Explanation:**

The Fundamental Theorem of Calculus ties [differentiation](../02_Single_Variable_Differentiation/01_derivative_definition.ipynb) and [integration](./01_antiderivative_and_indefinite_integral.ipynb) together as inverse processes. Part 1 says that differentiating an integral with a variable upper limit returns the integrand, so every continuous function has an antiderivative. Part 2 — the evaluation theorem — says a definite integral can be computed by evaluating *any* antiderivative at the endpoints and subtracting, replacing the [Riemann-sum limit](./02_definite_integral_and_riemann_sums.ipynb) with a single subtraction. This is what makes integrals practically computable.

**Properties:**
- Part 1 combined with the chain rule gives $\dfrac{d}{dx}\int_{a}^{g(x)} f(t)\,dt = f\big(g(x)\big)\,g'(x)$.
- Part 2 is the net change theorem: integrating a rate of change recovers the total change.

**Numerical Example:**

Recompute $\int_{0}^{2} x^2\,dx$ with Part 2 instead of Riemann sums. An antiderivative of $x^2$ is $F(x) = \tfrac{x^3}{3}$, so

$$
\int_{0}^{2} x^2\,dx = F(2) - F(0) = \frac{2^3}{3} - \frac{0^3}{3} = \frac{8}{3} \approx 2.667 ,
$$

matching the Riemann-sum limit exactly. For Part 1, with $G(x) = \int_{0}^{x} \cos t\,dt$, differentiating returns the integrand: $G'(x) = \cos x$.

```python
import sympy as sp

x, t = sp.symbols("x t")
f = x**2

antiderivative = sp.integrate(f, x)
evaluation = antiderivative.subs(x, 2) - antiderivative.subs(x, 0)
direct_definite = sp.integrate(f, (x, 0, 2))

variable_limit_integral = sp.integrate(sp.cos(t), (t, 0, x))
part1_derivative = sp.diff(variable_limit_integral, x)

print("F(x) =", antiderivative)
print("Part 2: F(2) - F(0) =", evaluation, "  sympy definite =", direct_definite)
print("G(x) = ∫_0^x cos t dt =", variable_limit_integral)
print("Part 1: d/dx G(x) =", part1_derivative)
```
