### 3.1.1. Optimization Problem


$$
\begin{aligned}
\min_{x \in X} \quad & f(x), \qquad X \subseteq \mathbb{R}^n, \\
x^* \in X \text{ is optimal}
&\Longleftrightarrow
f(x^*) \le f(x), \quad \forall x \in X.
\end{aligned}
$$


**Explanation:**

A mathematical optimization problem has three basic parts: a decision variable $x$, a set $X$ of admissible decisions, and a cost function $f$ that assigns a real number to each admissible decision. This model separates the question of what may be chosen from the question of how good each choice is. In this repository, the decision variable is usually a vector in a finite-dimensional [vector space](../../01_Foundations/01_Linear_Algebra/05_Geometrical_Aspects_of_Linear_Algebra/13_vector_space.ipynb), so the constraint set is a subset of $\mathbb{R}^n$. An optimal decision is one whose cost is no larger than the cost of any other admissible decision.

Read the display as a minimization problem over admissible choices. The [vector](../../01_Foundations/01_Linear_Algebra/02_Vector/01_vector_definition.ipynb) $x$ is the decision variable, $X$ is the feasible set, $f(x)$ is the cost, and $x^*$ is optimal when every feasible $x$ has cost at least $f(x^*)$.

This general definition is intentionally broad. If $X=\mathbb{R}^n$, the problem is unconstrained; if $X$ is described by equations or inequalities, the problem is constrained; if $X$ contains only discrete choices, different combinatorial ideas may be needed. The unconstrained optimization sequence specializes this broad model to unconstrained nonlinear programming, where the central problem is to minimize a differentiable cost over all of $\mathbb{R}^n$.

**Intuition:**

Suppose a small design study offers several admissible scalar decisions and a cost is computed for each one. The optimizer is the admissible decision with the smallest cost. The code cell demonstrates this finite version of the definition before the rest of the module moves to continuous unconstrained problems.


**Numerical Example:**

For a finite version of the definition, take
$$
X=\{-0.5,1.0,1.5,2.0\},\qquad f(x)=(x-1.5)^2+0.5.
$$

Evaluate the cost at every admissible decision:
$$
\begin{aligned}
f(-0.5)&=(-0.5-1.5)^2+0.5=4+0.5=4.5,\\
f(1.0)&=(1.0-1.5)^2+0.5=0.25+0.5=0.75,\\
f(1.5)&=(1.5-1.5)^2+0.5=0+0.5=0.5,\\
f(2.0)&=(2.0-1.5)^2+0.5=0.25+0.5=0.75.
\end{aligned}
$$

Since $0.5$ is the smallest value in the list, $x^*=1.5$ is optimal over this admissible set.


```python
import numpy as np

decision_candidates = np.array([-0.5, 1.0, 1.5, 2.0])
objective_values = (decision_candidates - 1.5) ** 2 + 0.5
best_index = np.argmin(objective_values)
optimal_decision = decision_candidates[best_index]
optimal_cost = objective_values[best_index]

print("decision_candidates =", decision_candidates)
print("objective_values =", objective_values)
print("optimal_decision =", optimal_decision)
print("optimal_cost =", optimal_cost)
```
