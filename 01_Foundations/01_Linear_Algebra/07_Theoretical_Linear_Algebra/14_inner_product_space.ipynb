### 1.1.7.14. Inner Product Space


$$
\langle \cdot, \cdot \rangle : V \times V \to \mathbb{R}
$$


**Explanation:**

An inner product space is a vector space equipped with a generalized dot product. Savov uses the inner product to transfer orthogonality, norm, and distance from Euclidean vectors to abstract vectors, so this notebook keeps those geometric consequences bundled into one concept. More specialized examples are moved to [generalized inner products](./15_generalized_inner_product.ipynb).

**Properties:**
- Symmetry
- Linearity in one argument
- Positive definiteness

**Example:**

In $\mathbb{R}^2$, the standard dot product defines an inner product space.


```python
import numpy as np

vector_u = np.array([3.0, 4.0])
vector_v = np.array([1.0, 2.0])
inner_product = float(np.dot(vector_u, vector_v))
norm_value = float(np.sqrt(np.dot(vector_u, vector_u)))

print("inner_product =", inner_product)
print("norm_value =", norm_value)
```
