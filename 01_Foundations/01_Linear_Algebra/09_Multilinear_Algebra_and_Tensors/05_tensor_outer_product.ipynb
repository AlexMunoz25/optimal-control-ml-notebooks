### 1.1.9.5. Tensor Outer Product

$$
C^{ij} = A^i B^j,
\qquad
C^{ik}{}_{jlm} = A^i{}_j B^k{}_{lm},
\qquad
\operatorname{rank}(C) = \operatorname{rank}(A) + \operatorname{rank}(B)
$$

**Explanation:**

The outer product multiplies tensor components without summing over any index. Every index from the input tensors remains in the result, so the ranks add. This extends [Matrix Outer Product](../03_Matrix/14_matrix_outer_product.ipynb) from vectors to arbitrary tensors.

**Example:**

If
$$
A^i =
\begin{bmatrix}
1 \\
2 \\
3
\end{bmatrix},
\qquad
B^j =
\begin{bmatrix}
4 \\
5 \\
6
\end{bmatrix},
$$
then
$$
A^i B^j =
\begin{bmatrix}
4 & 5 & 6 \\
8 & 10 & 12 \\
12 & 15 & 18
\end{bmatrix}.
$$

```python
import numpy as np

vector_a = np.array([1.0, 2.0, 3.0])
vector_b = np.array([4.0, 5.0, 6.0])
rank_2_outer_product = np.outer(vector_a, vector_b)

matrix_factor = np.array([[1.0, 0.0], [0.0, 2.0]])
rank_4_outer_product = np.einsum('ij,kl->ijkl', matrix_factor, matrix_factor)

print("rank_2_outer_product =\n", rank_2_outer_product)
print("rank_4_outer_product shape =", rank_4_outer_product.shape)
```
