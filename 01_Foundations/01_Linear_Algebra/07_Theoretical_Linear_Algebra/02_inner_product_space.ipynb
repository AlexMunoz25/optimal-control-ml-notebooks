### 1.1.7.2. Inner Product Space


$$
\langle \cdot, \cdot \rangle : V \times V \to \mathbb{R}
$$


**Explanation:**

An inner product space is a vector space equipped with a generalized dot product. This structure transfers orthogonality, norm, and distance from Euclidean vectors to abstract vectors, while more specialized examples appear in [generalized inner products](./15_generalized_inner_product.ipynb).

**Properties:**
- Symmetry
- Linearity in one argument
- Positive definiteness

**Example:**

In $\mathbb{R}^2$, the standard dot product defines an inner product space.


```python
import sympy as sp

vector_u = sp.Matrix([3, 4])
vector_v = sp.Matrix([1, 2])
inner_product = vector_u.dot(vector_v)
norm_value = vector_u.norm()

print("inner_product =", inner_product)
print("norm_value =", norm_value)
```
