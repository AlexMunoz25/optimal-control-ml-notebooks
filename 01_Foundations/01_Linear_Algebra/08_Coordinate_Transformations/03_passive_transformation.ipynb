### 1.1.8.3. Passive Transformation


$$
\left[ \mathbf{A} \right]_{B'} =
{}_{B'}\!\left[ \mathbb{1} \right]_B
\left[ \mathbf{A} \right]_B
$$


**Explanation:**

A passive transformation changes the coordinate system while keeping the geometric vector fixed. Only the component column changes, so this is the coordinate-side counterpart of the [active transformation](./04_active_transformation.ipynb). It is the natural interpretation of a [coordinate-system transformation](./01_coordinate_system_transformation.ipynb) when the frame, not the vector, is changed.

<p align="center">
  <img src="../../../Figures/01010801_passive_transformation_matrix.jpeg" alt="Passive transformation matrix" width="420">
</p>

**Example:**

When the axes rotate by $150^{\circ}$, the same vector receives new coordinates obtained from the passive transformation matrix.


```python
import numpy as np

rotation_angle = np.deg2rad(150.0)
passive_matrix = np.array(
    [
        [np.cos(rotation_angle), np.sin(rotation_angle)],
        [-np.sin(rotation_angle), np.cos(rotation_angle)],
    ]
)
vector_components = np.array([5.0, 3.0])
passive_components = passive_matrix @ vector_components

print("passive_components =", np.round(passive_components, 4))
```
