### 1.3.5.4. Directional Derivative

$$
D_{\mathbf{u}} f(\mathbf{x}) = \nabla f(\mathbf{x}) \cdot \mathbf{u}
= \|\nabla f(\mathbf{x})\|\,\cos\varphi, \qquad \|\mathbf{u}\| = 1.
$$

**Explanation:**

The directional derivative is the rate of change of $f$ along an arbitrary unit direction $\mathbf{u}$, generalizing the partial derivatives (which use the coordinate directions). It equals the [inner product](../../01_Linear_Algebra/02_Vector/04_vector_dot_product.ipynb) of the [gradient](./03_gradient.ipynb) with $\mathbf{u}$, so by the cosine form it is largest when $\mathbf{u}$ aligns with $\nabla f$ ($\varphi = 0$), zero when $\mathbf{u}$ is orthogonal to the gradient, and most negative opposite to it. This is the precise statement that the gradient is the steepest-ascent direction and $-\nabla f$ the steepest-descent direction used by [descent methods](../../../03_Optimization/09_Algorithms/01_descent_directions.ipynb).

**Properties:**
- $-\|\nabla f\| \le D_{\mathbf{u}} f \le \|\nabla f\|$, with the bounds attained along $\pm\nabla f / \|\nabla f\|$.
- A direction $\mathbf{u}$ with $D_{\mathbf{u}} f < 0$ is a descent direction for $f$.

**Intuition:**

The directional derivative is the slope of the surface along the chosen unit direction $\mathbf{u}$ at the point.

**Numerical Example:**

For $f(x,y) = x^2 + y^2$ at $(1, 2)$ we have $\nabla f(1,2) = [2, 4]^\top$. Along the unit direction $\mathbf{u} = \left[\tfrac35, \tfrac45\right]$,

$$
D_{\mathbf{u}} f(1,2) = \nabla f(1,2)\cdot\mathbf{u} = 2\cdot\tfrac35 + 4\cdot\tfrac45 = \tfrac{6}{5} + \tfrac{16}{5} = \tfrac{22}{5} = 4.4 .
$$

This is less than the maximal rate $\|\nabla f(1,2)\| = \sqrt{20} \approx 4.47$, attained only along the gradient direction itself.

```python
import sympy as sp

x, y = sp.symbols("x y")
f = x**2 + y**2

gradient = sp.Matrix([sp.diff(f, x), sp.diff(f, y)]).subs({x: 1, y: 2})
direction = sp.Matrix([sp.Rational(3, 5), sp.Rational(4, 5)])

directional_derivative = gradient.dot(direction)
maximal_rate = gradient.norm()

print("∇f(1,2)              =", gradient.T)
print("‖u‖                  =", direction.norm())
print("D_u f = ∇f · u       =", directional_derivative)
print("maximal rate ‖∇f‖    =", sp.N(maximal_rate, 6))
```
