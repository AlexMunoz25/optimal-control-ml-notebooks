### 3.9.16. Newton Method for Self-Concordant Functions

$$x^+=x-\frac{1}{1+\lambda(x)}\nabla^2f(x)^{-1}\nabla f(x)$$


**Explanation:**

Newton methods for self-concordant functions use the Newton decrement to choose damped steps before full steps are justified. The decrement measures local model reliability and predicted decrease, so feasibility is maintained while optimality is approached by driving the decrement down.

The update uses the usual Newton direction but scales it by $1/(1+\lambda(x))$. The quantity $\lambda(x)$ is the Newton decrement, so a larger decrement produces a more damped step.

**Properties:**
- Large decrement suggests damping.
- Small decrement indicates the fast Newton phase.
- The standard damped step scales by 1/(1+lambda).

**Intuition:**

The example computes a damped Newton step from a decrement.


**Numerical Example:**

Suppose
$$
\nabla f=2,\qquad \nabla^2f=4.
$$

The Newton step is
$$
\Delta x_{\rm nt}=-\frac{2}{4}=-\frac12.
$$

The decrement is
$$
\lambda=\sqrt{\frac{2^2}{4}}=1.
$$

The damped step is
$$
\frac{1}{1+\lambda}\Delta x_{\rm nt}=\frac12\left(-\frac12\right)=-\frac14.
$$


```python
import sympy as sp

gradient = sp.Integer(2)
hessian = sp.Integer(4)
newton_step = -gradient / hessian
decrement = sp.sqrt(gradient ** 2 / hessian)
damped_step = newton_step / (1 + decrement)

print("newton_step =", newton_step)
print("decrement =", decrement)
print("damped_step =", damped_step)
```
