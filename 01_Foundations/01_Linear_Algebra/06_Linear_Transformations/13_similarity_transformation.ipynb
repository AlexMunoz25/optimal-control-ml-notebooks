### 1.1.6.13. Similarity Transformation


$$
M = P N P^{-1}
$$


**Explanation:**

A similarity transformation is the square-matrix change-of-basis formula. Savov interprets the matrix $P$ as the adaptor that moves between coordinate systems, so similar matrices represent the same linear transformation in different bases. This notebook depends on the more general notebook on [change of basis for matrix representations](./12_change_of_basis_for_matrix_representations.ipynb).

**Example:**

If $N$ is diagonal and $P$ is invertible, then $M = P N P^{-1}$ describes the same transformation in the basis formed by the columns of $P$.


```python
import sympy as sp

matrix_n = sp.Matrix([[1, 0], [0, -1]])
matrix_p = sp.Matrix([[1, 1], [1, -1]])
matrix_m = sp.simplify(matrix_p * matrix_n * matrix_p.inv())

print("matrix_m =")
sp.pprint(matrix_m)
```
