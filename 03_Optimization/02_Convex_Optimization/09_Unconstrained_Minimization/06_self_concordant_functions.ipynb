### 3.2.9.6. Self-Concordant Functions


$$|f^{\prime\prime\prime}(x)|\leq2f^{\prime\prime}(x)^{3/2}$$


**Explanation:**

Self-concordance controls third derivatives using second derivatives, which is central to the analysis of logarithmic barriers and interior-point methods. Feasibility means staying in the domain interior; optimality is approached using local norms defined by curvature.

**Properties:**
- The logarithmic barrier is a standard self-concordant example.
- The defining inequality compares the third derivative with the second derivative.
- Self-concordance gives global Newton-step control.

**Intuition:**

The example verifies the scalar self-concordance inequality at one point.


**Numerical Example:**

Let
$$
f(x)=-\log x,\qquad x>0.
$$

Then
$$
f''(x)=\frac1{x^2},\qquad f'''(x)=-\frac2{x^3}.
$$

At $x=1$,
$$
|f'''(1)|=2.
$$

The right side of the self-concordance bound is
$$
2(f''(1))^{3/2}=2(1)^{3/2}=2.
$$

Therefore the inequality holds with equality at this point.


```python
import sympy as sp

point = sp.Integer(1)
second_derivative = 1 / point ** 2
third_derivative_abs = abs(-2 / point ** 3)
bound = 2 * second_derivative ** sp.Rational(3, 2)

third_derivative_abs, bound, third_derivative_abs <= bound
```
