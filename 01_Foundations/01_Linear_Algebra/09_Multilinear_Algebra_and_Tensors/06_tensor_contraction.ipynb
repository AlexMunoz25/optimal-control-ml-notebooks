### 1.1.9.6. Tensor Contraction

$$
D^j{}_l = C^{ij}{}_{il}
$$

$$
\operatorname{rank}(D) = \operatorname{rank}(C) - 2
$$

$$
\text{one upper index contracted with one lower index}
$$


**Explanation:**

- Contraction sets one upper index equal to one lower index and sums over that repeated index.
- Each contraction removes two slots, so the rank drops by two.
- The upper-lower pairing matters because it lets the Jacobian and inverse Jacobian collapse to a Kronecker delta.

**Connections:**

- The trace from the matrix module is the simplest tensor contraction.
- `np.einsum` and `tf.einsum` implement contractions directly through repeated indices.

**Example:**

- For $C^{ij}{}_{kl} = A^i{}_k B^j{}_l$, contracting $i$ with $k$ gives
$$
D^j{}_l = C^{ij}{}_{il} = (\operatorname{tr} A)\,B^j{}_l.
$$


```python
import numpy as np

matrix_a = np.array([[1.0, 2.0], [3.0, 4.0]])
matrix_b = np.array([[5.0, 6.0], [7.0, 8.0]])
rank_4_tensor = np.einsum('ik,jl->ijkl', matrix_a, matrix_b)

matrix_a_trace = np.einsum('ii->', matrix_a)
contracted_tensor = np.einsum('ijil->jl', rank_4_tensor)
expected_tensor = matrix_a_trace * matrix_b

print("matrix_a_trace =", matrix_a_trace)
print("contracted_tensor =\n", contracted_tensor)
print("expected_tensor =\n", expected_tensor)
```
