### 1.1.7.16. Gram-Schmidt Orthogonalization


$$
\mathbf{e}_j =
\mathbf{v}_j -
\sum_{i=1}^{j-1}
\frac{\langle \mathbf{v}_j, \hat{\mathbf{e}}_i \rangle}{\lVert \hat{\mathbf{e}}_i \rVert^2}
\hat{\mathbf{e}}_i
$$


**Explanation:**

Gram-Schmidt converts a basis into an orthogonal or orthonormal basis by removing previously accounted-for projection components. In an [inner product space](./14_inner_product_space.ipynb), those components are defined through [projection](../05_Geometrical_Aspects_of_Linear_Algebra/07_projection_onto_a_line.ipynb), and the same construction underlies [QR decomposition](./20_qr_decomposition.ipynb).

**Example:**

Applying Gram-Schmidt to the columns of a full-rank matrix produces the orthogonal factor used later in [QR decomposition](./20_qr_decomposition.ipynb).


```python
import sympy as sp

basis_vectors = [sp.Matrix([1, 1, 0]), sp.Matrix([1, 0, 1]), sp.Matrix([0, 1, 1])]
orthogonal_basis = []

for current_vector in basis_vectors:
    projection_sum = sum(
        (
            current_vector.dot(previous_vector) / previous_vector.dot(previous_vector) * previous_vector
            for previous_vector in orthogonal_basis
        ),
        sp.zeros(current_vector.rows, 1),
    )
    orthogonal_basis.append(sp.simplify(current_vector - projection_sum))

print("orthogonal_basis =", orthogonal_basis)
```
