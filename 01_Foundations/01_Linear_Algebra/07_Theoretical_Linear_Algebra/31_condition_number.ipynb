### 1.1.7.31. Condition Number


$$
\kappa(A) = \|A\|\,\|A^{-1}\|
$$

$$
\kappa_2(A) = \frac{\sigma_{\max}(A)}{\sigma_{\min}(A)}
$$

$$
Q \succ 0 \quad \Longrightarrow \quad \kappa_2(Q) = \frac{\lambda_{\max}(Q)}{\lambda_{\min}(Q)}
$$


**Explanation:**

The condition number measures how unevenly a nonsingular matrix stretches different directions, and how much solving a linear system can amplify perturbations. In the spectral norm, it is the ratio of the largest to smallest singular value from the [singular value decomposition](./28_singular_value_decomposition.ipynb). For a [positive definite matrix](./17_positive_definite_matrix.ipynb), the singular values match the positive [eigenvalues](./08_eigenvalue.ipynb), giving the eigenvalue ratio used in optimization rate analysis.

**Example:**

For

$$
Q =
\begin{bmatrix}
25 & 0 \\
0 & 1
\end{bmatrix},
$$

the eigenvalues are $25$ and $1$, so

$$
\kappa_2(Q) = \frac{25}{1} = 25.
$$


```python
import sympy as sp

matrix_q = sp.Matrix([[25, 0], [0, 1]])
eigenvalues = list(matrix_q.eigenvals().keys())
condition_number = max(eigenvalues) / min(eigenvalues)

print("eigenvalues =", eigenvalues)
print("condition_number =", condition_number)
```
