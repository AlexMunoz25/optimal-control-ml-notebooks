### 1.1.9.13. Vectors and One-Forms

$$
\mathbf{v} = v^i \mathbf{e}_i,
\qquad
\tilde{\omega} = \omega_i \mathbf{e}^{\,i},
\qquad
\langle \tilde{\omega}, \mathbf{v} \rangle = \omega_i v^i
$$

**Explanation:**

In the modern language emphasized at the end of Fleisch's chapter, vectors and one-forms are distinct rank-1 objects rather than two component descriptions of the same object. Vectors transform contravariantly and expand in the original basis, while one-forms transform covariantly and expand in the dual basis from [Dual Basis](../08_Coordinate_Transformations/06_dual_basis.ipynb). Their pairing is an invariant scalar, and the metric from [Index Raising and Lowering](./10_index_raising_lowering.ipynb) provides the map between a vector and its associated one-form.

**Example:**

Let

$$
\mathbf{v}=\begin{bmatrix}3\\3/2\end{bmatrix},
\qquad
\omega=\begin{bmatrix}2\\-1\end{bmatrix}.
$$

The one-form acts on the vector by pairing components:

$$
\omega(\mathbf{v})=2\cdot3+(-1)\cdot\frac32=6-\frac32=\frac92.
$$

With metric

$$
g=\begin{bmatrix}1&0\\0&4\end{bmatrix},
$$

the associated one-form is

$$
g\mathbf{v}=\begin{bmatrix}1&0\\0&4\end{bmatrix}\begin{bmatrix}3\\3/2\end{bmatrix}=\begin{bmatrix}3\\6\end{bmatrix}.
$$

The metric norm squared is

$$
\mathbf{v}^Tg\mathbf{v}=3\cdot3+\frac32\cdot6=9+9=18.
$$


```python
import sympy as sp

vector_components = sp.Matrix([3, sp.Rational(3, 2)])
one_form_components = sp.Matrix([2, -1])
metric_tensor = sp.Matrix([[1, 0], [0, 4]])

scalar_pairing = one_form_components.dot(vector_components)
associated_one_form = metric_tensor * vector_components
metric_norm_squared = (vector_components.T * metric_tensor * vector_components)[0]

print("scalar_pairing =", scalar_pairing)
print("associated_one_form =")
sp.pprint(associated_one_form.T)
print("metric_norm_squared =", metric_norm_squared)
```
