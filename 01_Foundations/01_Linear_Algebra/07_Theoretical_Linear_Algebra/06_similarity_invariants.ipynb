### 1.1.7.6. Similarity Invariants


$$
\det(A) = \prod_i \lambda_i,
\qquad
\operatorname{tr}(A) = \sum_i \lambda_i
$$


**Explanation:**

Savov identifies determinant and trace as quantities that do not change under similarity transformations. This notebook keeps that invariance as the concept and links back to [similarity transformations](../06_Linear_Transformations/13_similarity_transformation.ipynb) instead of rederiving the basis-change formula. The eigenvalue viewpoint explains why those quantities are basis independent.

**Example:**

For

$$
A =
\begin{bmatrix}
9 & -2 \\
-2 & 6
\end{bmatrix},
$$

the determinant is $50$ and the trace is $15$, matching the product and sum of the eigenvalues $5$ and $10$.


```python
import sympy as sp

matrix_a = sp.Matrix([[9, -2], [-2, 6]])
eigenvalues = list(matrix_a.eigenvals().keys())

print("determinant_value =", matrix_a.det())
print("trace_value =", matrix_a.trace())
print("product_of_eigenvalues =", sp.prod(eigenvalues))
print("sum_of_eigenvalues =", sum(eigenvalues))
```
