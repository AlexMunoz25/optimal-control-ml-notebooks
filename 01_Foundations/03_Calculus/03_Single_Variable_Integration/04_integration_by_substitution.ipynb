### 1.3.3.4. Integration by Substitution

$$
\int f\big(g(x)\big)\,g'(x)\,dx = \int f(u)\,du, \qquad u = g(x),\ du = g'(x)\,dx .
$$

**Explanation:**

Substitution is the [chain rule](../02_Single_Variable_Differentiation/03_chain_rule.ipynb) read backwards. When the integrand contains an inner function $g(x)$ multiplied by its own derivative $g'(x)$, the change of variable $u = g(x)$ collapses the composition into a simpler integral in $u$. For a definite integral the limits transform too, $x: a \to b$ becoming $u: g(a) \to g(b)$, which avoids converting back. It is the single most used integration technique and the one-dimensional ancestor of the [change-of-variables Jacobian](../06_Multivariable_Integration_and_Vector_Calculus/03_change_of_variables_and_jacobian_determinant.ipynb) in multiple integrals.

**Properties:**
- Definite form: $\displaystyle\int_{a}^{b} f\big(g(x)\big)g'(x)\,dx = \int_{g(a)}^{g(b)} f(u)\,du$.
- Recognizing the pattern "inner function and its derivative" is the key step.

**Numerical Example:**

Evaluate $\int_{0}^{1} 2x\,e^{x^{2}}\,dx$. Let $u = x^2$, so $du = 2x\,dx$ and the limits become $u: 0 \to 1$:

$$
\int_{0}^{1} 2x\,e^{x^2}\,dx = \int_{0}^{1} e^{u}\,du = e^{u}\big|_{0}^{1} = e^{1} - e^{0} = e - 1 \approx 1.718 .
$$

The factor $2x$ was exactly the derivative of the inner function $x^2$, which is what made the substitution close the integral.

```python
import sympy as sp

x, u = sp.symbols("x u")
integrand = 2 * x * sp.exp(x**2)

inner_function = x**2
transformed_integrand = sp.exp(u)
substituted_result = sp.integrate(transformed_integrand, (u, inner_function.subs(x, 0), inner_function.subs(x, 1)))
direct_result = sp.integrate(integrand, (x, 0, 1))

print("after u = x^2 :  ∫_0^1 e^u du =", substituted_result, "=", sp.N(substituted_result))
print("direct sympy  :", direct_result, "=", sp.N(direct_result))
```
