### 1.3.6.4. Vector Fields and Gradient Fields

$$
\mathbf{F}(x, y) = P(x, y)\,\mathbf{i} + Q(x, y)\,\mathbf{j};
\qquad
\mathbf{F} = \nabla f \iff \mathbf{F}\ \text{is conservative (a gradient field)} .
$$

**Explanation:**

A vector field assigns a vector to each point of space — a velocity, a force, or the [gradient](../05_Multivariable_Differential_Calculus/03_gradient.ipynb) of a scalar potential. A field that *is* the gradient of some scalar function $f$ is called conservative, and $f$ is its potential; conservative fields are exactly the "downhill arrows" of a landscape. On a simply connected domain, $\mathbf{F} = (P, Q)$ is conservative iff the cross-partials match, $P_y = Q_x$. Gradient fields are the picture of every objective in optimization, and the potential/conservative structure is what makes energy-based and physics-informed models well posed.

**Properties:**
- Conservative test (2-D, simply connected): $\dfrac{\partial P}{\partial y} = \dfrac{\partial Q}{\partial x}$.
- A gradient field is curl-free; the rotational field $\mathbf{F} = (-y, x)$ fails the test and has no potential.

**Intuition:**

A vector field attaches a vector to every point; the radial field shown points outward and is the gradient field of $\tfrac12(x^2+y^2)$.

**Numerical Example:**

Test two fields. For $\mathbf{F} = (2x, 2y)$: $P = 2x$, $Q = 2y$, so $P_y = 0 = Q_x$. It passes, and indeed $\mathbf{F} = \nabla f$ with potential $f(x,y) = x^2 + y^2$.

For the rotational field $\mathbf{G} = (-y, x)$: $P = -y$, $Q = x$, so

$$
P_y = -1, \qquad Q_x = 1, \qquad P_y \neq Q_x .
$$

$\mathbf{G}$ is **not** conservative — no scalar potential exists; its arrows circulate rather than point downhill.

```python
import sympy as sp

x, y = sp.symbols("x y")
gradient_field = (2 * x, 2 * y)
rotational_field = (-y, x)

def is_conservative(field):
    P, Q = field
    return sp.simplify(sp.diff(P, y) - sp.diff(Q, x)) == 0

def potential_function(field):
    P, Q = field
    partial_potential = sp.integrate(P, x)
    remaining_in_y = sp.simplify(Q - sp.diff(partial_potential, y))
    return partial_potential + sp.integrate(remaining_in_y, y)

potential = potential_function(gradient_field)

print("F = (2x, 2y):  P_y - Q_x =", sp.diff(gradient_field[0], y) - sp.diff(gradient_field[1], x),
      "-> conservative:", is_conservative(gradient_field))
print("  recovered potential f =", potential, "(+ const), since ∇f = F")
print("G = (-y, x):   P_y - Q_x =", sp.diff(rotational_field[0], y) - sp.diff(rotational_field[1], x),
      "-> conservative:", is_conservative(rotational_field))
```
