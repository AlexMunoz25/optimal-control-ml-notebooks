### 1.4.2.4. Linear First-Order Equations

$$
\frac{dy}{dx} + P(x)\,y = f(x), \qquad
\mu(x) = e^{\int P(x)\,dx}, \qquad
\frac{d}{dx}\big[\mu(x)\,y\big] = \mu(x)\,f(x) .
$$

**Explanation:**

A first-order equation is **linear** when it can be put in the **standard form** $y' + P(x)y = f(x)$. Multiplying by the **integrating factor** $\mu(x) = e^{\int P\,dx}$ turns the left side into the exact derivative of a product, $\frac{d}{dx}[\mu y]$, so both sides can be integrated directly. The procedure is: (i) put the equation in standard form; (ii) identify $P(x)$ and form $\mu = e^{\int P\,dx}$; (iii) multiply through, recognizing the left side as $\frac{d}{dx}[\mu y]$; (iv) integrate and solve for $y$. This integrating factor is a special case of the [integrating factors](./06_integrating_factors.ipynb) that make an equation exact, and the final integration often uses [integration by parts](../../03_Calculus/03_Single_Variable_Integration/05_integration_by_parts.ipynb).

**Numerical Example:**

Solve the IVP $\dfrac{dy}{dx} + y = x$, $y(0) = 4$.

**(i)** It is already in standard form with $P(x) = 1$ and $f(x) = x$.

**(ii)** The integrating factor is

$$
\mu(x) = e^{\int 1\,dx} = e^{x}.
$$

**(iii)** Multiply the equation by $e^x$; the left side collapses to a product derivative:

$$
e^{x}y' + e^{x}y = x e^{x} \;\Longrightarrow\; \frac{d}{dx}\big[e^{x}y\big] = x e^{x}.
$$

**(iv)** Integrate both sides, using integration by parts on the right ($\int x e^x\,dx = x e^x - e^x$):

$$
e^{x}y = x e^{x} - e^{x} + C \;\Longrightarrow\; y = x - 1 + C e^{-x}.
$$

Apply $y(0) = 4$: $\;4 = 0 - 1 + C \Rightarrow C = 5$. Hence

$$
y = x - 1 + 5e^{-x}.
$$

The term $5e^{-x}$ is a **transient**: it decays to $0$, so every solution approaches the line $y = x - 1$ as $x \to \infty$.

```python
import sympy as sp

x = sp.symbols("x")
y = sp.Function("y")

P = 1
integrating_factor = sp.exp(sp.integrate(P, x))
right_side_integral = sp.integrate(integrating_factor * x, x)

general_solution = sp.dsolve(sp.Eq(y(x).diff(x) + y(x), x), y(x))
particular_solution = sp.dsolve(sp.Eq(y(x).diff(x) + y(x), x), y(x), ics={y(0): 4})

print("integrating factor mu =", integrating_factor)
print("∫ x e^x dx           =", right_side_integral)
print("general solution      :", general_solution)
print("with y(0) = 4         :", particular_solution)
```
