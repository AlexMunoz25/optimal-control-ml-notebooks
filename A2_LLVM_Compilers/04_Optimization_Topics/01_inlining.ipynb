### A2.4.1. Inlining

> *Inlining replaces a function call with the body of the called function, eliminating call overhead and exposing optimization opportunities across the former call boundary.*

**Explanation:**

**Inlining** is one of the most impactful compiler optimizations. By replacing a call with the callee's body, the optimizer can:

- Eliminate call/return overhead.
- Specialize the callee's code with the caller's context (constant arguments, known types).
- Enable further optimizations across the inlined boundary (constant propagation, dead code elimination).

**Cost Model:**

LLVM's inliner uses a cost/benefit model:

- **Cost** â€” estimated increase in code size (instruction count after inlining).
- **Threshold** â€” a configurable limit. If cost > threshold, the call is not inlined.
- **Bonuses** â€” applied for small functions, single-call-site functions, functions called in hot loops.

**Properties:**

- `always_inline` attribute forces inlining regardless of cost.
- `noinline` attribute prevents inlining.
- Recursive calls are not inlined (would cause infinite expansion).
- The inliner operates at CGSCC level in the new pass manager.

**Example:**

```llvm
; Before inlining:
define i32 @add(i32 %a, i32 %b) { %r = add i32 %a, %b  ret i32 %r }
define i32 @caller() { %v = call i32 @add(i32 3, i32 4)  ret i32 %v }

; After inlining + constant folding:
define i32 @caller() { ret i32 7 }
```

````python
inline_decisions = [
    ("small function (5 instrs)", 5, 225, "INLINE"),
    ("medium function (100 instrs)", 100, 225, "INLINE"),
    ("large function (300 instrs)", 300, 225, "NO â€” cost > threshold"),
    ("always_inline attribute", 500, 225, "INLINE â€” forced"),
    ("noinline attribute", 5, 225, "NO â€” blocked"),
]

print("Inlining Decision Examples:")
print(f"{'Scenario':<32} {'Cost':<8} {'Threshold':<12} {'Decision'}")
print("-" * 72)
for scenario, cost, threshold, decision in inline_decisions:
    print(f"{scenario:<32} {cost:<8} {threshold:<12} {decision}")
````
