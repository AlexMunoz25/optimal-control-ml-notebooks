### 1.4.1.1. Definitions and Terminology

$$
F\big(x, y, y', \ldots, y^{(n)}\big) = 0
\qquad\xrightarrow{\text{linear}}\qquad
a_n(x)\,y^{(n)} + a_{n-1}(x)\,y^{(n-1)} + \cdots + a_1(x)\,y' + a_0(x)\,y = g(x).
$$

**Explanation:**

A **differential equation (DE)** is an equation containing the [derivatives](../../03_Calculus/02_Single_Variable_Differentiation/01_derivative_definition.ipynb) of one or more unknown functions with respect to one or more independent variables. Differential equations are classified three ways. By **type**: an *ordinary* differential equation (ODE) involves derivatives with respect to a single independent variable, while a *partial* differential equation involves several. By **order**: the order of the highest derivative present (so $y'' + 5(y')^3 - 4y = e^x$ is second order). By **linearity**: the equation is **linear** when it can be written in the form above — the unknown $y$ and every [derivative](../../03_Calculus/02_Single_Variable_Differentiation/06_higher_order_derivatives.ipynb) appear to the first power and each coefficient depends at most on $x$; otherwise it is **nonlinear**.

**Properties:**
- The two tests for linearity: (i) $y, y', \ldots, y^{(n)}$ each appear to the first degree; (ii) the coefficients depend at most on the independent variable $x$.
- Nonlinear terms in the dependent variable — such as $y^2$, $y\,y'$, $\sin y$, or $e^{y'}$ — make an equation nonlinear.

**Numerical Example:**

Classify four equations by order and linearity.

1. $y' + 5y = e^x$. Highest derivative $y'$, so **order 1**; $y$ and $y'$ appear to the first power with coefficients depending only on $x$, so **linear**.
2. $y'' - y' + 6y = 0$. Highest derivative $y''$, so **order 2**; first-degree terms, constant coefficients, so **linear**.
3. $(1 - y)\,y' + 2y = e^x$. Highest derivative $y'$, so **order 1**; but the coefficient $(1 - y)$ of $y'$ depends on the dependent variable $y$, violating test (ii), so **nonlinear**.
4. $y'' + \sin y = 0$ (the pendulum equation). Highest derivative $y''$, so **order 2**; the term $\sin y$ is a nonlinear function of $y$, so **nonlinear**.

So order counts derivatives, while linearity is decided by *how* $y$ and its derivatives enter.

```python
import sympy as sp

x = sp.symbols("x")
y = sp.Function("y")

equations = {
    "y' + 5y = e^x": sp.Eq(y(x).diff(x) + 5 * y(x), sp.exp(x)),
    "y'' - y' + 6y = 0": sp.Eq(y(x).diff(x, 2) - y(x).diff(x) + 6 * y(x), 0),
    "(1 - y) y' + 2y = e^x": sp.Eq((1 - y(x)) * y(x).diff(x) + 2 * y(x), sp.exp(x)),
    "y'' + sin(y) = 0": sp.Eq(y(x).diff(x, 2) + sp.sin(y(x)), 0),
}

for label, ode in equations.items():
    order = sp.ode_order(ode, y(x))
    hints = sp.classify_ode(ode, y(x))
    is_linear = any("linear" in hint and "nonlinear" not in hint for hint in hints)
    print(f"{label:24} order = {order}   linear = {is_linear}")
```
