### 1.1.8.1. Coordinate-System Transformation


$$
\begin{aligned}
A^{x'} &= A^x \cos(\alpha_{11}) + A^y \cos(\alpha_{12}), \\
A^{y'} &= A^x \cos(\alpha_{21}) + A^y \cos(\alpha_{22}), \\
\begin{bmatrix}
A^{x'} \\
A^{y'}
\end{bmatrix}
&=
\begin{bmatrix}
\cos(\alpha_{11}) & \cos(\alpha_{12}) \\
\cos(\alpha_{21}) & \cos(\alpha_{22})
\end{bmatrix}
\begin{bmatrix}
A^x \\
A^y
\end{bmatrix}
=
\begin{bmatrix}
\cos \theta & \sin \theta \\
-\sin \theta & \cos \theta
\end{bmatrix}
\begin{bmatrix}
A^x \\
A^y
\end{bmatrix}
\end{aligned}
$$


**Explanation:**

Rotating the coordinate axes does not change the geometric vector; it only changes the numerical components used to describe that vector. Because the new axes point partly along the old directions, each new component is a weighted linear combination of the original components. The cosine factors are the direction cosines between the original and rotated axes.

**Example:**

If the original components are
$$
\begin{bmatrix}
A^x \\
A^y
\end{bmatrix}
=
\begin{bmatrix}
5 \\
3
\end{bmatrix}
$$
and the axes rotate counterclockwise by $150^{\circ}$, then
$$
\begin{bmatrix}
A^{x'} \\
A^{y'}
\end{bmatrix}
=
\begin{bmatrix}
\cos 150^{\circ} & \cos 60^{\circ} \\
\cos 240^{\circ} & \cos 150^{\circ}
\end{bmatrix}
\begin{bmatrix}
5 \\
3
\end{bmatrix}
=
\begin{bmatrix}
-2.83 \\
-5.10
\end{bmatrix}.
$$


```python
import numpy as np

rotation_angle_degrees = 150.0
rotation_angle = np.deg2rad(rotation_angle_degrees)
passive_transformation_matrix = np.array(
    [
        [np.cos(rotation_angle), np.sin(rotation_angle)],
        [-np.sin(rotation_angle), np.cos(rotation_angle)],
    ]
)
original_vector_components = np.array([5.0, 3.0])
transformed_vector_components = passive_transformation_matrix @ original_vector_components

print("passive_transformation_matrix =")
print(np.round(passive_transformation_matrix, 4))
print("transformed_vector_components =", np.round(transformed_vector_components, 4))
```
