### 3.4.21. Duality with Generalized Inequalities


$$L(x,\lambda)=f_0(x)+\langle\lambda,F(x)\rangle,\qquad\lambda\succeq_{K^*}0$$


**Explanation:**

Duality with [generalized inequalities](../02_Convex_Analysis/09_generalized_inequalities_and_proper_cones.ipynb) uses multipliers from the [dual cone](../02_Convex_Analysis/11_dual_cones_and_dual_generalized_inequalities.ipynb). These multipliers scalarize [cone-valued](../02_Convex_Analysis/03_cones_and_conic_combinations.ipynb) residuals in the Lagrangian while preserving the order needed for weak duality.

The residual $F(x)$ takes values ordered by a cone. The multiplier $\lambda$ lies in the [dual cone](../02_Convex_Analysis/11_dual_cones_and_dual_generalized_inequalities.ipynb), and the pairing $\langle\lambda,F(x)\rangle$ turns the cone-valued residual into a scalar Lagrangian term.

**Properties:**
- The cone defines the meaning of inequality.
- For the nonnegative orthant, every component must be nonnegative.
- This order supports vector optimization and conic duality.

**Intuition:**

The example checks cone membership by inspecting each component of a difference vector.


**Numerical Example:**

Let
$$
x=\begin{bmatrix}1\\2\end{bmatrix},\quad
y=\begin{bmatrix}3\\5\end{bmatrix},\quad K=\mathbf{R}_+^2.
$$

The order test is $x\preceq_K y$ if $y-x\in K$. Compute
$$
y-x=\begin{bmatrix}3-1\\5-2\end{bmatrix}=\begin{bmatrix}2\\3\end{bmatrix}.
$$

Both components are nonnegative, so $x\preceq_K y$.


```python
import sympy as sp

point_x = sp.Matrix([1, 2])
point_y = sp.Matrix([3, 5])
difference = point_y - point_x
nonnegative = [entry >= 0 for entry in difference]

print("difference =", difference)
print("nonnegative =", nonnegative)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

point_x = ca.DM([1, 2])
point_y = ca.DM([3, 5])
difference = point_y - point_x
nonnegative = [bool(difference[index] >= 0) for index in range(difference.numel())]

print("difference =", difference)
print("nonnegative =", nonnegative)
```
