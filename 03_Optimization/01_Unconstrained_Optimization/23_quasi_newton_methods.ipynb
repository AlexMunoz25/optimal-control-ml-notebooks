### 3.1.23. Quasi-Newton Methods


$$
p^k=x^{k+1}-x^k,
\qquad
q^k=\nabla f(x^{k+1})-\nabla f(x^k),
\qquad
D^{k+1}q^k=p^k.
$$


**Explanation:**

Quasi-Newton methods try to obtain much of Newton's local speed without computing the Hessian explicitly. After a step has been taken, the change in position $p^k$ and the change in gradient $q^k$ contain curvature information. The secant relation $D^{k+1}q^k=p^k$ asks the inverse-Hessian approximation to be consistent with that observed curvature. If $D^k$ remains positive definite, then $-D^k\nabla f(x^k)$ is a descent direction.

The central idea is learning curvature from motion. Steepest descent uses no curvature, Newton computes curvature directly, and quasi-Newton methods update an approximation from successive gradients. Bertsekas places these methods after conjugate gradient because both seek better directions than steepest descent, but they make different storage and computation tradeoffs.

**Example:**

For a quadratic function, the gradient changes exactly according to the Hessian times the step, so the secant relation is exact. For a nonquadratic function, it is a local approximation that becomes more informative near a solution. The code cell checks the exact secant behavior on a small quadratic model.


```python
import numpy as np

Q = np.array([[3.0, 1.0], [1.0, 2.0]])
previous_point = np.array([0.0, 0.0])
next_point = np.array([1.0, -1.0])
previous_gradient = Q @ previous_point
next_gradient = Q @ next_point
position_increment = next_point - previous_point
gradient_increment = next_gradient - previous_gradient
secant_residual = gradient_increment - Q @ position_increment

secant_residual
```
