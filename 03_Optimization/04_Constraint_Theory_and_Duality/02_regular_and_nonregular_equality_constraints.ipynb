### 3.4.2. Regular and Nonregular Equality Constraints

$$
V(x^*)=\{y\mid \nabla h_i(x^*)^T y=0,\ i=1,\ldots,m\}
$$

$$
\nabla f(x^*) + \sum_{i=1}^{m} \lambda_i^* \nabla h_i(x^*) = 0
$$

**Explanation:**

A feasible point is regular when the active equality constraint gradients are linearly independent, matching the local [basis](../../01_Foundations/01_Linear_Algebra/05_Geometrical_Aspects_of_Linear_Algebra/10_basis.ipynb)-style geometry of the constraint normals. In that case, the [subspace](../../01_Foundations/01_Linear_Algebra/05_Geometrical_Aspects_of_Linear_Algebra/15_vector_subspace.ipynb) of first-order feasible variations has the expected dimension, and the Lagrange multiplier equation correctly expresses the absence of first-order descent along the constraint surface. If regularity fails, the linearized feasible variations may be larger than the true local feasible set, and multipliers can fail to exist even at a global minimum.

**Intuition:**

<img src="../../Figures/030402_berts_fig_3_1_2_no_lagrange_multipliers.jpeg" alt="Bertsekas Figure 3.1.2: a nonregular equality-constrained problem with no Lagrange multipliers" style="width: 60%; height: auto;">

The figure shows two equality constraints whose intersection is only the origin. Their gradients at the origin point in the same horizontal direction, so the linearized constraint system incorrectly allows vertical first-order variations.

**Numerical Example:**

Consider the problem shown in the figure:

$$
\min f(x)=x_1+x_2
$$

subject to

$$
h_1(x)=(x_1-1)^2+x_2^2-1=0,
\qquad
h_2(x)=(x_1-2)^2+x_2^2-4=0.
$$

At

$$
x^*=\begin{bmatrix}0\\0\end{bmatrix},
$$

the constraints are satisfied because

$$
h_1(x^*)=(-1)^2+0-1=0,
\qquad
h_2(x^*)=(-2)^2+0-4=0.
$$

The gradients are

$$
\nabla f(x^*)=\begin{bmatrix}1\\1\end{bmatrix},
\quad
\nabla h_1(x^*)=\begin{bmatrix}-2\\0\end{bmatrix},
\quad
\nabla h_2(x^*)=\begin{bmatrix}-4\\0\end{bmatrix}.
$$

The multiplier equation would require

$$
\begin{bmatrix}1\\1\end{bmatrix}
+\lambda_1\begin{bmatrix}-2\\0\end{bmatrix}
+\lambda_2\begin{bmatrix}-4\\0\end{bmatrix}
=\begin{bmatrix}0\\0\end{bmatrix}.
$$

The second coordinate becomes

$$
1+0+0=0,
$$

which is impossible. The local minimum exists, but no Lagrange multipliers exist because the constraint gradients are dependent and do not span the needed normal direction.

```python
import sympy as sp

x_1, x_2, lambda_1, lambda_2 = sp.symbols('x_1 x_2 lambda_1 lambda_2')

cost = x_1 + x_2
constraint_1 = (x_1 - 1)**2 + x_2**2 - 1
constraint_2 = (x_1 - 2)**2 + x_2**2 - 4
candidate = {x_1: 0, x_2: 0}

gradient_cost = sp.Matrix([sp.diff(cost, x_1), sp.diff(cost, x_2)]).subs(candidate)
gradient_constraints = sp.Matrix([
    [sp.diff(constraint_1, x_1), sp.diff(constraint_2, x_1)],
    [sp.diff(constraint_1, x_2), sp.diff(constraint_2, x_2)],
]).subs(candidate)
stationarity_equation = gradient_cost + gradient_constraints * sp.Matrix([lambda_1, lambda_2])
solutions = sp.solve(stationarity_equation, [lambda_1, lambda_2], dict=True)

result = {
    'gradient_cost': gradient_cost,
    'gradient_constraints': gradient_constraints,
    'constraint_gradient_rank': gradient_constraints.rank(),
    'multiplier_solutions': solutions,
}

for label, value in result.items():
    print(f"{label} =", value)
```
