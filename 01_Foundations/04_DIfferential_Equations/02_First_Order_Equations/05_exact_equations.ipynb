### 1.4.2.5. Exact Equations

$$
M(x, y)\,dx + N(x, y)\,dy = 0 \ \text{ is exact} \iff \frac{\partial M}{\partial y} = \frac{\partial N}{\partial x},
$$

$$
\text{then } \exists\, f(x,y): \ \frac{\partial f}{\partial x} = M,\ \frac{\partial f}{\partial y} = N, \quad \text{solution } f(x, y) = c .
$$

**Explanation:**

A differential form $M\,dx + N\,dy$ is **exact** when it is the total differential $df$ of some function $f(x,y)$. The criterion is the equality of the cross [partial derivatives](../../03_Calculus/05_Multivariable_Differential_Calculus/02_partial_derivatives.ipynb), $\partial M/\partial y = \partial N/\partial x$ — the same condition that makes a planar [vector field conservative](../../03_Calculus/06_Multivariable_Integration_and_Vector_Calculus/04_vector_fields_and_gradient_fields.ipynb), with $f$ playing the role of the potential. The solution is then the implicit family $f(x,y) = c$, recovered by integrating $M$ in $x$, adding an unknown $g(y)$, and fixing $g$ by matching $\partial f/\partial y = N$.

**Properties:**
- The criterion comes from equality of mixed partials: $f_{xy} = f_{yx}$.
- The implicit solution $f(x,y) = c$ is a level curve of the potential $f$.

**Numerical Example:**

Solve $2xy\,dx + (x^2 - 1)\,dy = 0$.

**Step 1 — test exactness.** With $M = 2xy$ and $N = x^2 - 1$,

$$
\frac{\partial M}{\partial y} = 2x = \frac{\partial N}{\partial x}, \quad\text{so the equation is exact.}
$$

**Step 2 — integrate $M$ in $x$** (holding $y$ fixed), introducing an unknown $g(y)$:

$$
f(x, y) = \int 2xy\,dx = x^2 y + g(y).
$$

**Step 3 — match $\partial f/\partial y = N$:**

$$
\frac{\partial f}{\partial y} = x^2 + g'(y) = x^2 - 1 \;\Rightarrow\; g'(y) = -1 \;\Rightarrow\; g(y) = -y .
$$

**Step 4 — assemble and write $f = c$:**

$$
f(x, y) = x^2 y - y \;\Rightarrow\; x^2 y - y = c \;\Rightarrow\; y = \frac{c}{x^2 - 1},
$$

valid on any interval not containing $x = \pm 1$.

```python
import sympy as sp

x, y = sp.symbols("x y")
M = 2 * x * y
N = x**2 - 1

is_exact = sp.simplify(sp.diff(M, y) - sp.diff(N, x)) == 0

potential = sp.integrate(M, x)
g_prime = sp.simplify(N - sp.diff(potential, y))
potential = potential + sp.integrate(g_prime, y)

print("M_y =", sp.diff(M, y), "  N_x =", sp.diff(N, x), "  exact:", is_exact)
print("f(x,y) =", potential)
print("implicit solution: f(x,y) = c  ->  x^2 y - y = c")
```

```python
import numpy as np
import matplotlib.pyplot as plt

grid = np.linspace(-3, 3, 400)
mesh_x, mesh_y = np.meshgrid(grid, grid)
potential_values = mesh_x**2 * mesh_y - mesh_y

plt.figure(figsize=(7, 5))
contour = plt.contour(mesh_x, mesh_y, potential_values, levels=[-6, -3, -1, 0, 1, 3, 6], cmap="viridis")
plt.clabel(contour, inline=True, fontsize=8)
plt.axvline(1, color="gray", linestyle="--", alpha=0.5); plt.axvline(-1, color="gray", linestyle="--", alpha=0.5)
plt.title(r"Solution family $x^2 y - y = c$ (level curves of the potential)")
plt.xlabel("x"); plt.ylabel("y"); plt.ylim(-3, 3); plt.tight_layout()
plt.show()
```
