### 1.1.5.16. Matrix Fundamental Spaces


$$
\mathcal{C}(A), \qquad \mathcal{R}(A), \qquad \mathcal{N}(A), \qquad \mathcal{N}(A^{\mathsf{T}})
$$


**Explanation:**

The four fundamental spaces of a matrix are the column space, row space, null space, and left null space. Together they describe how a matrix acts on both inputs and outputs, and they connect directly to the transformation viewpoints of [image space](../06_Linear_Transformations/02_image_space.ipynb) and [kernel](../06_Linear_Transformations/03_kernel.ipynb).

**Properties:**
- $\mathcal{R}(A)$ is orthogonal to $\mathcal{N}(A)$.
- $\mathcal{C}(A)$ is orthogonal to $\mathcal{N}(A^{\mathsf{T}})$.

**Example:**

For

$$
A =
\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 9
\end{bmatrix},
$$

the rank is $2$, so the nullity is $1$.


```python
import sympy as sp

matrix_a = sp.Matrix([[1, 2, 3], [4, 5, 6], [7, 8, 9]])
rank_value = matrix_a.rank()
nullity_value = matrix_a.cols - rank_value

print("rank_value =", rank_value)
print("nullity_value =", nullity_value)
print("column_space_dimension =", len(matrix_a.columnspace()))
print("null_space_dimension =", len(matrix_a.nullspace()))
```
