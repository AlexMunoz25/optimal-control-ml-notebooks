### 3.2.9.4. Newton Step, Newton Decrement, and Affine Invariance


$$\Delta x_{\rm nt}=-\nabla^2f(x)^{-1}\nabla f(x)$$


**Explanation:**

The Newton step minimizes the local quadratic approximation of a twice differentiable [convex function](../03_Convex_Functions/01_convex_function_definition_and_examples.ipynb). The Hessian supplies the local curvature metric, and the Newton decrement measures predicted decrease. Feasibility is unconstrained here; optimality is approached by solving a linear system.

The Newton step multiplies the gradient by the inverse Hessian and changes sign. The Hessian inverse is a [matrix inverse](../../../01_Foundations/01_Linear_Algebra/03_Matrix/11_matrix_inverse.ipynb), so the step rescales the gradient by local curvature before moving downhill.

**Properties:**
- The step solves a Hessian-gradient system.
- The decrement is a curvature-scaled gradient [norm](../../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb).
- Newton geometry explains affine invariance.

**Intuition:**

The figure above shows local quadratic geometry; the arithmetic solves one local Newton system.

**Numerical Example:**

The figure above shows local quadratic geometry; the arithmetic solves one local Newton system.

Let
$$
\nabla f=\begin{bmatrix}2\\4\end{bmatrix},\quad
\nabla^2 f=\begin{bmatrix}2&0\\0&4\end{bmatrix}.
$$

The Newton equation is
$$
\nabla^2 f\,\Delta x=-\nabla f.
$$

Thus
$$
2\Delta x_1=-2 \Rightarrow \Delta x_1=-1,
\qquad
4\Delta x_2=-4 \Rightarrow \Delta x_2=-1.
$$

The squared decrement is
$$
\frac{2^2}{2}+\frac{4^2}{4}=2+4=6.
$$


```python
import sympy as sp

gradient = sp.Matrix([2, 4])
hessian = sp.diag(2, 4)
step = -hessian.inv() * gradient

decrement_squared = (gradient.T * hessian.inv() * gradient)[0]
step, decrement_squared
```
