### 1.4.8.3. Repeated and Complex Eigenvalues

$$
\textbf{complex } \lambda = \alpha \pm i\beta,\ \mathbf{K} = \mathbf{B}_1 + i\mathbf{B}_2:
\quad
\mathbf{X}_1 = (\mathbf{B}_1\cos\beta t - \mathbf{B}_2\sin\beta t)e^{\alpha t},\ \
\mathbf{X}_2 = (\mathbf{B}_2\cos\beta t + \mathbf{B}_1\sin\beta t)e^{\alpha t}.
$$

**Explanation:**

The [eigenvalue method](./02_homogeneous_systems_eigenvalue_method.ipynb) needs two refinements. For a **repeated** eigenvalue with too few eigenvectors, a second solution uses a *generalized eigenvector*: $\mathbf{X}_2 = (\mathbf{K}t + \mathbf{P})e^{\lambda t}$ where $(A - \lambda I)\mathbf{P} = \mathbf{K}$. For a **complex** eigenvalue $\alpha \pm i\beta$ with eigenvector $\mathbf{K} = \mathbf{B}_1 + i\mathbf{B}_2$, Euler's formula converts the complex exponentials into the two real solutions above — a decaying/growing oscillation. Complex eigenvalues mean the origin is a **spiral** (or a **center** when $\alpha = 0$), the rotational modes that pervade vibrating and oscillatory control systems.

**Properties:**
- $\mathbf{B}_1 = \operatorname{Re}(\mathbf{K})$ and $\mathbf{B}_2 = \operatorname{Im}(\mathbf{K})$ are the real and imaginary parts of the complex eigenvector.
- $\alpha < 0$: stable spiral; $\alpha > 0$: unstable spiral; $\alpha = 0$: center (closed orbits).

**Numerical Example:**

Solve the IVP $\mathbf{X}' = \begin{pmatrix} 2 & 8 \\ -1 & -2 \end{pmatrix}\mathbf{X}$, $\mathbf{X}(0) = \begin{pmatrix} 2 \\ -1 \end{pmatrix}$.

**Eigenvalues.** $\det(A - \lambda I) = \lambda^2 + 4 = 0 \Rightarrow \lambda = \pm 2i$, so $\alpha = 0$, $\beta = 2$ (a center).

**Complex eigenvector.** For $\lambda = 2i$: $(2-2i)k_1 + 8k_2 = 0$ gives $k_1 = -(2+2i)k_2$; choosing $k_2 = -1$,

$$
\mathbf{K} = \begin{pmatrix} 2+2i \\ -1 \end{pmatrix} = \underbrace{\begin{pmatrix} 2 \\ -1 \end{pmatrix}}_{\mathbf{B}_1} + i\underbrace{\begin{pmatrix} 2 \\ 0 \end{pmatrix}}_{\mathbf{B}_2} .
$$

**Real solutions** (with $\alpha = 0$, $\beta = 2$):

$$
\mathbf{X} = c_1\!\left(\begin{pmatrix} 2 \\ -1 \end{pmatrix}\cos 2t - \begin{pmatrix} 2 \\ 0 \end{pmatrix}\sin 2t\right) + c_2\!\left(\begin{pmatrix} 2 \\ 0 \end{pmatrix}\cos 2t + \begin{pmatrix} 2 \\ -1 \end{pmatrix}\sin 2t\right).
$$

**Apply** $\mathbf{X}(0) = (2, -1)$: this gives $2c_1 + 2c_2 = 2$ and $-c_1 = -1$, so $c_1 = 1$, $c_2 = 0$, and

$$
\mathbf{X} = \begin{pmatrix} 2\cos 2t - 2\sin 2t \\ -\cos 2t \end{pmatrix},
$$

a closed elliptical orbit (center) through $(2, -1)$.

```python
import sympy as sp

t = sp.symbols("t")
x, y = sp.symbols("x y", cls=sp.Function)

A = sp.Matrix([[2, 8], [-1, -2]])
eigenvalues = list(A.eigenvals().keys())

system = [sp.Eq(x(t).diff(t), 2 * x(t) + 8 * y(t)), sp.Eq(y(t).diff(t), -x(t) - 2 * y(t))]
solution = sp.dsolve(system, ics={x(0): 2, y(0): -1})

print("eigenvalues:", eigenvalues, "(pure imaginary -> center)")
for equation in solution:
    print(sp.simplify(equation))
```
