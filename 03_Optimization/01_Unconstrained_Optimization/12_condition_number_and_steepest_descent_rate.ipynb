### 3.1.12. Condition Number and Steepest Descent Rate


$$
\kappa(Q)=\frac{M}{m},
\qquad
\frac{\|x^{k+1}\|}{\|x^k\|}
\le \max\{|1-\alpha^k m|,|1-\alpha^k M|\},
\qquad
\alpha^*=\frac{2}{M+m}.
$$


**Explanation:**

For a positive definite quadratic, the rate of steepest descent is governed by the smallest and largest [eigenvalues](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/08_eigenvalue.ipynb) of the Hessian matrix $Q$. Their ratio is the [condition number](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/31_condition_number.ipynb), and a large ratio means the level sets are elongated. In such a geometry, the negative gradient tends to point across the narrow direction of the valley, producing the familiar zig-zag behavior. The contraction bound in the definition cell shows this dependence explicitly through the two endpoint eigenvalues $m$ and $M$.

This is one of the main places where linear algebra becomes optimization intuition. Spectral properties are not abstract decorations; they determine how quickly a first-order method can reduce error on the local quadratic model. This analysis motivates scaling, preconditioning, Newton's method, and quasi-Newton methods.

**Intuition:**

The first figure shows the stepsize that minimizes the worst-case contraction bound. The second shows that the bound is sharp for certain starting points, so slow convergence is not just a proof artifact. The Kantorovich figure connects the line-search rate result to eigenvalue geometry. The code cell computes the optimal constant stepsize and the associated contraction factor from $m$ and $M$.

**Numerical Example:**

The rate-bound figure depends only on the endpoint eigenvalues. Take
$$
m=1,
\qquad M=30.
$$

The condition number is
$$
\kappa(Q)=\frac{M}{m}=\frac{30}{1}=30.
$$

The constant stepsize that minimizes the worst endpoint contraction is
$$
\alpha^*=\frac{2}{M+m}=\frac{2}{30+1}=\frac{2}{31}\approx 0.0645.
$$

The associated contraction bound is
$$
\frac{\kappa-1}{\kappa+1}=\frac{30-1}{30+1}=\frac{29}{31}\approx 0.9355.
$$

A bound close to $1$ means that an ill-conditioned quadratic can reduce the error very slowly under steepest descent.


```python
smallest_eigenvalue = 1.0
largest_eigenvalue = 30.0
optimal_stepsize = 2.0 / (smallest_eigenvalue + largest_eigenvalue)
condition_number = largest_eigenvalue / smallest_eigenvalue
contraction_bound = (condition_number - 1.0) / (condition_number + 1.0)

optimal_stepsize, condition_number, contraction_bound
```
