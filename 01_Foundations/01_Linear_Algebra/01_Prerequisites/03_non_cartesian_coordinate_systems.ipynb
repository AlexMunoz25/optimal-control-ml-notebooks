### 1.1.1.3. Non-Cartesian Coordinate Systems

$$
\textbf{Polar:} \quad x = r\cos\theta, \quad y = r\sin\theta
$$

$$
\textbf{Cylindrical:} \quad x = r\cos\phi, \quad y = r\sin\phi, \quad z = z
$$

$$
\textbf{Spherical:} \quad x = r\sin\theta\cos\phi, \quad y = r\sin\theta\sin\phi, \quad z = r\cos\theta
$$

**Explanation:**

While Cartesian coordinates describe positions using perpendicular axes, many problems are more naturally expressed in **curvilinear coordinate systems** where the coordinate lines follow curves.

In **polar coordinates**, a point is specified by distance $r$ from the origin and angle $\theta$ from the positive $x$-axis. The unit vectors $\hat{r}$ and $\hat{\theta}$ point in the directions of increasing $r$ and $\theta$, and their directions change with location.

       alt="2-D rectangular and polar coordinates"
       width="650">
</p>

In three dimensions, **cylindrical coordinates** $(r, \phi, z)$ extend polar coordinates by adding a vertical axis $z$:

**Spherical coordinates** $(r, \theta, \phi)$ use radial distance $r$, polar angle $\theta$ (from the $z$-axis), and azimuthal angle $\phi$ (from the $x$-axis in the $xy$-plane):

</p>

Compare this notebook with [The Cartesian Plane](./02_cartesian_plane.ipynb), where the coordinate lines are perpendicular and straight. This viewpoint is useful before [Vectors](../02_Vector/01_vector_definition.ipynb), where coordinates become vector components.

**Example:**

Take the Cartesian point

$$
(x,y)=(1,\sqrt{3}).
$$

The polar radius is

$$
r=\sqrt{x^2+y^2}=\sqrt{1^2+(\sqrt{3})^2}=\sqrt{1+3}=2.
$$

The angle satisfies

$$
\cos\theta=\frac{x}{r}=\frac{1}{2}, \qquad
\sin\theta=\frac{y}{r}=\frac{\sqrt{3}}{2},
$$

so

$$
\theta=\frac{\pi}{3}=60^\circ.
$$

Converting back gives

$$
x=r\cos\theta=2\cdot\frac{1}{2}=1,
\qquad
y=r\sin\theta=2\cdot\frac{\sqrt{3}}{2}=\sqrt{3}.
$$

Thus $(x,y)=(1,\sqrt{3})$ and $(r,\theta)=(2,\pi/3)$ describe the same point.


```python
import sympy as sp

cartesian_x = sp.Integer(1)
cartesian_y = sp.sqrt(3)

radius = sp.sqrt(cartesian_x**2 + cartesian_y**2)
angle = sp.atan2(cartesian_y, cartesian_x)

back_to_x = sp.simplify(radius * sp.cos(angle))
back_to_y = sp.simplify(radius * sp.sin(angle))

print("Cartesian:", (cartesian_x, cartesian_y))
print("Polar: r =", radius, ", θ =", angle, "rad =", sp.deg(angle), "°")
print("Back to Cartesian:", (back_to_x, back_to_y))
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

angles = np.linspace(0, 2 * np.pi, 200)
radii = [1.0, 2.0, 3.0]

fig, (cartesian_ax, polar_ax) = plt.subplots(1, 2, figsize=(12, 5))

for current_radius in radii:
    circle_x = current_radius * np.cos(angles)
    circle_y = current_radius * np.sin(angles)
    cartesian_ax.plot(circle_x, circle_y, label=f"r = {current_radius}")

point_x_exact = sp.Integer(1)
point_y_exact = sp.sqrt(3)
point_radius = sp.sqrt(point_x_exact**2 + point_y_exact**2)
point_angle = sp.atan2(point_y_exact, point_x_exact)
point_x = float(point_x_exact)
point_y = float(point_y_exact)
point_angle_float = float(point_angle)
point_radius_float = float(point_radius)

cartesian_ax.scatter([point_x], [point_y], color="red", s=100, zorder=5,
                     label=f"P = ({point_x:.0f}, {point_y:.2f})")
cartesian_ax.plot([0, point_x], [0, point_y], "r--", linewidth=1.5)
cartesian_ax.set_xlabel("x")
cartesian_ax.set_ylabel("y")
cartesian_ax.set_title("Cartesian View")
cartesian_ax.set_aspect("equal")
cartesian_ax.legend()
cartesian_ax.grid(True, alpha=0.3)

polar_ax = plt.subplot(122, projection="polar")

print("cartesian_point =", (point_x_exact, point_y_exact))
print("polar_radius =", point_radius)
print("polar_angle =", point_angle)
print("polar_angle_degrees =", sp.deg(point_angle))

polar_ax.plot([0, point_angle_float], [0, point_radius_float], "r-o", linewidth=2,
              label=f"P = (r={point_radius_float:.1f}, θ={np.degrees(point_angle_float):.0f}°)")
polar_ax.set_title("Polar View", pad=20)
polar_ax.legend(loc="lower right")

plt.tight_layout()
plt.show()
```
