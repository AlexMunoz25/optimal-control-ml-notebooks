### 1.1.6.9. Projection Matrix Representation


$$
M_{\Pi_{\hat{\mathbf{a}}}} = \hat{\mathbf{a}} \hat{\mathbf{a}}^{\mathsf{T}}
$$


**Explanation:**

A projection matrix is the matrix form of the geometric projection defined earlier in [projection onto a line](../05_Geometrical_Aspects_of_Linear_Algebra/07_projection_onto_a_line.ipynb). It is a concrete example of the probing procedure for matrix representation, and the outer-product formula shows how directional geometry becomes matrix algebra.

**Example:**

For the line $y = x$ with unit direction

$$
\hat{\mathbf{a}} =
\frac{1}{\sqrt{2}}
\begin{bmatrix}
1 \\
1
\end{bmatrix},
$$

the projection matrix is

$$
M_{\Pi_{\hat{\mathbf{a}}}} =
\frac{1}{2}
\begin{bmatrix}
1 & 1 \\
1 & 1
\end{bmatrix}.
$$


```python
import sympy as sp

unit_direction = sp.Matrix([1, 1]) / sp.sqrt(2)
projection_matrix = sp.simplify(unit_direction * unit_direction.T)

print("projection_matrix =")
sp.pprint(projection_matrix)
```
