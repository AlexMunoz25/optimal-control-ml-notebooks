### 3.3.12. Convex Function Definition and Examples

$$f(\theta x+(1-\theta)y)\leq\theta f(x)+(1-\theta)f(y)$$


**Explanation:**

A convex function lies below the chord connecting any two points in its domain. This single inequality explains why local minima are global and why convex objectives fit naturally with convex feasible sets.

The points $x$ and $y$ lie in the domain of $f$, and $\theta$ is a weight between zero and one. The left side evaluates $f$ at the weighted average of the points; the right side takes the same weighted average of the two function values.

**Properties:**
- The input average is compared with the average of function values.
- Quadratics with nonnegative curvature are canonical convex examples.
- Jensen inequalities extend the same idea to many points.

**Intuition:**

The Boyd and Vandenberghe figure shows one chord for a convex function. The Tordesillas screenshot contrasts convex and nonconvex functions in one and higher dimensions: in the convex cases, the segment between two function values stays above the graph; in the nonconvex cases, some segment passes below part of the graph, so the convexity inequality fails.

**Numerical Example:**

The chord figures above show why the inequality compares the value at an averaged input against the averaged function values.

Let $f(x)=x^2$, $x=0$, $y=2$, and $\theta=1/4$. The averaged input is
$$
\theta x+(1-\theta)y=\frac14\cdot0+\frac34\cdot2=\frac32.
$$

The left side is
$$
f(3/2)=(3/2)^2=9/4.
$$

The right side is
$$
\frac14 f(0)+\frac34 f(2)=\frac14\cdot0+\frac34\cdot4=3.
$$

Since $9/4\leq3$, the convexity inequality holds.


```python
import sympy as sp

theta = sp.Rational(1, 4)
point_x = sp.Integer(0)
point_y = sp.Integer(2)
midpoint = theta * point_x + (1 - theta) * point_y
left_side = midpoint ** 2
right_side = theta * point_x ** 2 + (1 - theta) * point_y ** 2

print("left_side =", left_side)
print("right_side =", right_side)
print("left_side <= right_side =", left_side <= right_side)
```
