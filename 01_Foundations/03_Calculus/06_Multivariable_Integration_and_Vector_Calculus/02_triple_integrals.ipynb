### 1.3.6.2. Triple Integrals

$$
\iiint_{E} f(x, y, z)\,dV
= \int_{a}^{b}\!\int_{c}^{d}\!\int_{p}^{q} f(x, y, z)\,dz\,dy\,dx .
$$

**Explanation:**

A triple integral accumulates a function over a three-dimensional solid $E$, evaluated as three nested [iterated integrals](./01_double_and_iterated_integrals.ipynb). With $f \equiv 1$ it returns the volume of $E$; with $f$ a mass density it returns total mass, and weighting by position gives centres of mass. In probability and statistical physics the same construction integrates a density over a 3-D state space, and it is the spatial integral that the [divergence theorem](./07_divergence_theorem_and_integration_by_parts.ipynb) connects to surface flux.

**Properties:**
- Volume of $E$: $V = \iiint_E 1\,dV$.
- Fubini extends to three (and more) variables when $f$ is continuous on $E$.

**Numerical Example:**

Integrate $f(x,y,z) = x + y + z$ over the unit cube $E = [0,1]^3$. Innermost in $z$:

$$
\int_{0}^{1}(x + y + z)\,dz = (x + y) + \tfrac12 .
$$

Then in $y$: $\int_0^1\big(x + y + \tfrac12\big)\,dy = x + \tfrac12 + \tfrac12 = x + 1$. Finally in $x$:

$$
\int_{0}^{1}(x + 1)\,dx = \tfrac12 + 1 = \frac{3}{2} .
$$

So $\iiint_E (x+y+z)\,dV = \tfrac32$, while $\iiint_E 1\,dV = 1$ recovers the cube's volume.

```python
import sympy as sp

x, y, z = sp.symbols("x y z")
f = x + y + z

triple_integral = sp.integrate(sp.integrate(sp.integrate(f, (z, 0, 1)), (y, 0, 1)), (x, 0, 1))
cube_volume = sp.integrate(sp.integrate(sp.integrate(1, (z, 0, 1)), (y, 0, 1)), (x, 0, 1))

print("∭_cube (x+y+z) dV =", triple_integral)
print("∭_cube 1 dV (volume) =", cube_volume)
```
