### 3.7.3. Robust Optimization

$$
\min_{\mathbf{x}}\; \mathbf{c}^\top \mathbf{x}
\quad\text{subject to}\quad
\mathbf{a}^\top \mathbf{x} \le b \;\; \forall\, \mathbf{a}\in\mathcal{U},
$$
with uncertainty set $\mathcal{U}$.

**Explanation:**

Robust optimization protects against uncertainty by requiring the constraints to hold for *every* realization of the uncertain data inside a chosen uncertainty set $\mathcal{U}$ — a worst-case rather than probabilistic guarantee. For a box uncertainty set $\mathbf{a}\in[\bar{\mathbf{a}}-\hat{\mathbf{a}},\,\bar{\mathbf{a}}+\hat{\mathbf{a}}]$, the semi-infinite constraint collapses to the single deterministic *robust counterpart* $\bar{\mathbf{a}}^\top\mathbf{x} + \hat{\mathbf{a}}^\top|\mathbf{x}| \le b$. This tightens the feasible set relative to the nominal problem, trading optimality for immunity to data variation.

![Box and ellipsoidal uncertainty sets](../../Figures/030703_robust_uncertainty_set.png)

**Numerical Example:**

$$
\max_{x_1,x_2\ge 0}\; x_1 + x_2
\quad\text{s.t.}\quad a_1 x_1 + a_2 x_2 \le 1 \;\; \forall\, a_i\in[0.8,\,1.2].
$$

Because $\mathbf{x}\ge 0$, the worst case takes each $a_i$ at its upper value $1.2$, giving the robust counterpart
$$
1.2\,(x_1 + x_2) \le 1 \;\Rightarrow\; x_1 + x_2 \le 0.833.
$$
The robust optimum has value $0.833$, below the nominal ($a_i=1$) optimum of $1$: the $0.167$ gap is the "price of robustness" buying feasibility for every coefficient in the set.

```python
import sympy as sp

nominal, half_width = sp.Rational(1), sp.Rational(2, 10)
worst_case_coefficient = nominal + half_width
robust_bound = sp.Rational(1) / worst_case_coefficient

print("worst-case coefficient =", float(worst_case_coefficient))
print("robust optimum x_1 + x_2 =", round(float(robust_bound), 3))
print("nominal optimum x_1 + x_2 =", 1.0, " price of robustness =", round(1 - float(robust_bound), 3))
```

**Equivalent scipy.linprog implementation:**

```python
import numpy as np
from scipy.optimize import linprog

worst_case = np.array([[1.2, 1.2]])
result = linprog(c=[-1.0, -1.0], A_ub=worst_case, b_ub=[1.0], bounds=[(0, None), (0, None)])

print("robust optimum x_1 + x_2 =", round(-result.fun, 3))
print("robust solution =", result.x.round(3))
```
