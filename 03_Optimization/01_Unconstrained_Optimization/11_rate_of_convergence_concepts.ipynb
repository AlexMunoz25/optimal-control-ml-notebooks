### 3.1.11. Rate of Convergence Concepts


$$
\begin{aligned}
e(x^k) &\le q\beta^k &&\text{linear convergence}, \\
\limsup_{k\to\infty}\frac{e(x^{k+1})}{e(x^k)} &= 0 &&\text{superlinear convergence}, \\
e(x^{k+1}) &= O(e(x^k)^2) &&\text{quadratic convergence}.
\end{aligned}
$$


**Explanation:**

Convergence alone does not say whether an algorithm is useful. A method may approach a stationary point but require so many iterations that it is impractical. Rate of convergence is therefore introduced by comparing an error sequence with standard sequences, especially geometric sequences. The error can be the distance to the limit point, the cost difference, or another nonnegative measure that vanishes at the solution.

Linear convergence means that the tail of the error behaves no worse than a fixed geometric contraction. Superlinear convergence means the contraction factor itself tends to zero, and quadratic convergence means the next error is proportional to the square of the current error. These distinctions prepare the reader for the central contrast in unconstrained optimization: steepest descent is simple but often linearly slow, while Newton-like methods can be much faster near a nonsingular minimizer.

**Intuition:**

If two methods both converge, the one with the faster asymptotic rate can become dramatically more accurate after only a few additional iterations. The code cell compares representative linear, superlinear, and quadratic error sequences, not as a theorem about a specific method, but as a numerical picture of the definitions.


**Numerical Example:**

Start three error sequences at simple values and compare their first few terms. A linearly convergent model with ratio $0.8$ gives
$$
e_0=1,
\qquad e_1=0.8e_0=0.8,
\qquad e_2=0.8e_1=0.64.
$$

A superlinear model $e_k=0.8^{2^k}$ gives
$$
e_0=0.8,
\qquad e_1=0.8^2=0.64,
\qquad e_2=0.8^4=0.4096,
\qquad e_3=0.8^8\approx 0.1678.
$$

Its successive ratios shrink:
$$
\frac{e_1}{e_0}=0.8,
\qquad
\frac{e_2}{e_1}=0.64,
\qquad
\frac{e_3}{e_2}=0.4096.
$$

A quadratically convergent model with $e_{k+1}=e_k^2$ and $e_0=0.5$ gives
$$
e_1=(0.5)^2=0.25,
\qquad e_2=(0.25)^2=0.0625,
\qquad e_3=(0.0625)^2=0.00390625.
$$

The linear ratios stay fixed:
$$
\frac{e_2}{e_1}=\frac{0.64}{0.8}=0.8.
$$

The quadratic ratios shrink even faster:
$$
\frac{0.25}{0.5}=0.5,
\qquad
\frac{0.0625}{0.25}=0.25,
\qquad
\frac{0.00390625}{0.0625}=0.0625.
$$

These numerical patterns distinguish linear, superlinear, and quadratic convergence.


```python
linear_errors = [0.8**iteration for iteration in range(8)]
superlinear_errors = [0.8**(2**iteration) for iteration in range(8)]
quadratic_errors = [0.5]

for iteration in range(7):
    quadratic_errors.append(quadratic_errors[-1] ** 2)

linear_errors, superlinear_errors, quadratic_errors
```
