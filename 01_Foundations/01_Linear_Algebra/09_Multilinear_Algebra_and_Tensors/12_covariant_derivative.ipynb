### 1.1.9.12. Covariant Derivative

$$
A^i{}_{;j} \equiv \frac{\partial A^i}{\partial x^j} + A^k \Gamma^i_{kj}
$$

$$
A_{i;j} \equiv \frac{\partial A_i}{\partial x^j} - A_k \Gamma^k_{ij}
$$

$$
A^{ij}{}_{;k} = \frac{\partial A^{ij}}{\partial x^k} + A^{lj} \Gamma^i_{lk} + A^{il} \Gamma^j_{lk}
$$

**Explanation:**

- Ordinary partial derivatives differentiate only tensor components.
- In curvilinear coordinates that is not enough, because the basis itself changes with position.
- The covariant derivative adds one Christoffel term for each upper index and subtracts one for each lower index.
- The result gains one extra lower index that records the differentiation direction.

**Connections:**

- In Cartesian coordinates every Christoffel symbol is zero, so the covariant derivative reduces to the usual partial derivative.
- The same idea appears whenever derivatives must respect geometry rather than raw coordinates.

**Example:**

- In cylindrical coordinates,
$$
A^r{}_{;\phi} = \frac{\partial A^r}{\partial \phi} - r A^\phi.
$$
- The extra term appears because the $\phi$ basis direction rotates as $\phi$ changes.


```python
import sympy as sp

radial_coordinate, angular_coordinate = sp.symbols('r phi', positive=True)
vector_component_radial, vector_component_angular = sp.symbols('A_r A_phi')

christoffel_radial_phiphi = -radial_coordinate
christoffel_angular_rphi = 1 / radial_coordinate

covariant_derivative_radial_wrt_phi = (
    sp.diff(vector_component_radial, angular_coordinate)
    + vector_component_angular * christoffel_radial_phiphi
)
covariant_derivative_angular_wrt_phi = (
    sp.diff(vector_component_angular, angular_coordinate)
    + vector_component_radial * christoffel_angular_rphi
)

print("A^r_{;phi} =", covariant_derivative_radial_wrt_phi)
print("A^phi_{;phi} =", covariant_derivative_angular_wrt_phi)
```
