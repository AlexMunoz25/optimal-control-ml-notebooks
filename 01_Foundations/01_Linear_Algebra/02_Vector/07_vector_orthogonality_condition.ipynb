### 1.1.2.7. Orthogonality Condition

$$
\mathbf{u} \cdot \mathbf{v} = 0
$$

**Explanation:**

Two vectors are orthogonal if their dot product is zero.
This condition is critical in linear algebra and ML for understanding independence and basis construction.

Depends on the [Dot Product](./04_vector_dot_product.ipynb). The same row-column dot-product viewpoint appears in [Matrix Multiplication](../03_Matrix/05_matrix_multiplication.ipynb).

**Example:**

If

$$
\mathbf{u} =
\begin{bmatrix}
1 \\
2
\end{bmatrix},
\quad
\mathbf{v} =
\begin{bmatrix}
-2 \\
1
\end{bmatrix},
$$

then

$$
\mathbf{u} \cdot \mathbf{v} = 1 \cdot (-2) + 2 \cdot 1 = 0,
$$

confirming orthogonality.


```python
import sympy as sp

vector_u = sp.Matrix([1, 2])
vector_v = sp.Matrix([-2, 1])

dot_product = vector_u.dot(vector_v)
is_orthogonal = dot_product == 0

print("u · v =", dot_product)
print("Orthogonal:", is_orthogonal)
```
