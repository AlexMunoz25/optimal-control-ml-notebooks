### 3.1.13. Scaling and Preconditioning


$$
x = Sy,
\qquad
x^{k+1}=x^k-\alpha^kD^k\nabla f(x^k),
\qquad
D^k = SS.
$$


**Explanation:**

Scaling changes the coordinates in which the optimization problem is viewed. A problem can be difficult for steepest descent simply because the variables are measured in incompatible units, causing the Hessian to have a large [condition number](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/31_condition_number.ipynb). By setting $x=Sy$, the same optimization problem is expressed in coordinates where the level sets may be much less elongated. In algorithmic form, this appears as a [positive definite matrix](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/17_positive_definite_matrix.ipynb) $D^k$ multiplying the gradient, and the goal is to make the scaled Hessian better conditioned.

The substitution $x=Sy$ changes coordinates with a scaling matrix $S$. In the iteration, $D^k$ plays the role of a preconditioner: it changes the geometry of the gradient step through a [matrix-vector product](../../01_Foundations/01_Linear_Algebra/03_Matrix/04_matrix_vector_multiplication.ipynb) before the stepsize is applied.

This point matters because poor scaling is common in engineering models. A control variable measured in volts, a state measured in meters, and a parameter measured in millions of dollars can all influence the same cost at very different numerical scales. The [condition number](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/31_condition_number.ipynb) and [change of basis](../../01_Foundations/01_Linear_Algebra/05_Geometrical_Aspects_of_Linear_Algebra/12_change_of_basis.ipynb) viewpoints explain why rescaling can make a simple method behave much better.

**Intuition:**

For a diagonal quadratic, choosing the coordinate scaling from the diagonal curvature can turn a poorly conditioned Hessian into one close to the identity. The code cell demonstrates this by comparing the condition number before and after a diagonal scaling. This is the simplest version of the preconditioning idea used later for conjugate gradient and Newton-like methods.


**Numerical Example:**

Let an unscaled quadratic have Hessian
$$
Q=\begin{bmatrix}100&0\\0&1\end{bmatrix}.
$$

Its condition number is
$$
\kappa(Q)=\frac{100}{1}=100.
$$

Choose the coordinate scaling
$$
S=\begin{bmatrix}0.1&0\\0&1\end{bmatrix}.
$$

The scaled Hessian is
$$
SQS=
\begin{bmatrix}0.1&0\\0&1\end{bmatrix}
\begin{bmatrix}100&0\\0&1\end{bmatrix}
\begin{bmatrix}0.1&0\\0&1\end{bmatrix}
=
\begin{bmatrix}1&0\\0&1\end{bmatrix}.
$$

Thus the condition number changes from $100$ to $1$, showing how scaling can turn an elongated quadratic into a spherical one.


```python
import sympy as sp

Q = sp.diag(100, 1)
scaling_matrix = sp.diag(sp.Rational(1, 10), 1)
scaled_hessian = scaling_matrix * Q * scaling_matrix

def condition_number(matrix):
    eigenvalues = sorted(matrix.eigenvals(), key=float)
    return eigenvalues[-1] / eigenvalues[0]

print("condition number of Q =", condition_number(Q))
print("condition number of scaled Hessian =", condition_number(scaled_hessian))
```
