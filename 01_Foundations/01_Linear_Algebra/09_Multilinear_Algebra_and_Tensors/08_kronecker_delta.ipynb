### 1.1.9.8. Kronecker Delta

$$
\delta^i{}_j =
\begin{cases}
1 & i = j \\
0 & i \neq j
\end{cases},
\qquad
\delta^i{}_j A^j = A^i,
\qquad
\delta^i{}_i = n
$$

**Explanation:**

The Kronecker delta is the mixed identity tensor. It appears whenever a direct Jacobian and an inverse Jacobian contract, so it is the algebraic reason contractions preserve tensor character. This is the tensor version of the identity matrix from the matrix notebooks, and in Cartesian orthonormal coordinates it matches the metric numerically.

**Example:**

If
$$
A^j =
\begin{bmatrix}
2 \\
5 \\
7
\end{bmatrix},
$$
then
$$
\delta^i{}_j A^j = A^i =
\begin{bmatrix}
2 \\
5 \\
7
\end{bmatrix},
\qquad
\delta^i{}_i = 3.
$$

```python
import numpy as np

kronecker_delta_tensor = np.eye(3)
vector_components = np.array([2.0, 5.0, 7.0])

identity_action = np.einsum('ij,j->i', kronecker_delta_tensor, vector_components)
self_contraction = np.einsum('ii->', kronecker_delta_tensor)

print("kronecker_delta_tensor =\n", kronecker_delta_tensor)
print("identity_action =", identity_action)
print("self_contraction =", self_contraction)
```
