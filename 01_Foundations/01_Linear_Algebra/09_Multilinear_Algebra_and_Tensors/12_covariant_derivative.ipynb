### 1.1.9.12. Covariant Derivative

$$
A^i{}_{;j} = \frac{\partial A^i}{\partial x^j} + A^k \Gamma^i_{kj},
\qquad
A_{i;j} = \frac{\partial A_i}{\partial x^j} - A_k \Gamma^k_{ij},
\qquad
C^i{}_{j;k} = \frac{\partial C^i{}_j}{\partial x^k} + C^l{}_j \Gamma^i_{lk} - C^i{}_l \Gamma^l_{jk}
$$

**Explanation:**

An ordinary partial derivative differentiates only tensor components, so in curvilinear coordinates it misses the change in the basis. Using the Christoffel symbols from [Christoffel Symbols](./11_christoffel_symbols.ipynb), the covariant derivative adds one correction term for each upper index and subtracts one for each lower index, ensuring that the result is again a tensor. In Cartesian coordinates every Christoffel symbol is zero, so the covariant derivative reduces to the usual partial derivative.

**Example:**

In cylindrical coordinates,
$$
A^r{}_{;\phi} = \frac{\partial A^r}{\partial \phi} - r A^\phi,
\qquad
A^\phi{}_{;\phi} = \frac{\partial A^\phi}{\partial \phi} + \frac{1}{r} A^r.
$$
The extra terms appear because the $r$- and $\phi$-basis directions rotate as $\phi$ changes.

```python
import sympy as sp

radial_coordinate, angular_coordinate = sp.symbols('r phi', positive=True)
radial_component = sp.Function('A_r')(angular_coordinate)
angular_component = sp.Function('A_phi')(angular_coordinate)

covariant_derivative_radial = (
    sp.diff(radial_component, angular_coordinate)
    - radial_coordinate * angular_component
)
covariant_derivative_angular = (
    sp.diff(angular_component, angular_coordinate)
    + radial_component / radial_coordinate
)

print("A^r_{;phi} =", covariant_derivative_radial)
print("A^phi_{;phi} =", covariant_derivative_angular)
```
