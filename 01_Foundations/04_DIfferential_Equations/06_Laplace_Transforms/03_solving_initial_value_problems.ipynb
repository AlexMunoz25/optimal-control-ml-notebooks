### 1.4.6.3. Solving Initial-Value Problems

$$
\mathcal{L}\{y'\} = sY(s) - y(0), \qquad
\mathcal{L}\{y''\} = s^2 Y(s) - s\,y(0) - y'(0) .
$$

**Explanation:**

Because the transform of a [derivative](../../03_Calculus/02_Single_Variable_Differentiation/01_derivative_definition.ipynb) is $sY(s)$ minus initial data, transforming a linear ODE replaces it with an **algebraic equation** for $Y(s)$ that already contains the initial conditions. The three-step method is: (1) transform the whole equation, (2) solve algebraically for $Y(s)$, (3) take the [inverse transform](./02_inverse_laplace_transform.ipynb) (via partial fractions) to get $y(t)$. Unlike the [undetermined-coefficients](../04_Higher_Order_Linear_Equations/04_undetermined_coefficients.ipynb) route, it builds the initial conditions in from the start, so no separate step is needed to fix the constants — exactly why it is the workhorse for linear, time-invariant systems in control.

**Properties:**
- For an $n$th-order IVP, the transform produces $P(s)Y(s) = (\text{initial-data polynomial}) + G(s)$, where $P$ is the characteristic polynomial.
- The method handles discontinuous and impulsive forcing that the classical methods handle awkwardly.

**Numerical Example:**

Solve $y'' - 3y' + 2y = e^{-4t}$, $y(0) = 1$, $y'(0) = 5$.

**Step 1 — transform**, using $\mathcal{L}\{y''\} = s^2Y - sy(0) - y'(0)$ and $\mathcal{L}\{y'\} = sY - y(0)$:

$$
\big(s^2Y - s - 5\big) - 3\big(sY - 1\big) + 2Y = \frac{1}{s+4} .
$$

**Step 2 — solve for $Y$.** Collecting, $(s^2 - 3s + 2)Y = s + 2 + \dfrac{1}{s+4}$, so

$$
Y(s) = \frac{s+2}{(s-1)(s-2)} + \frac{1}{(s-1)(s-2)(s+4)} = \frac{s^2 + 6s + 9}{(s-1)(s-2)(s+4)} .
$$

**Step 3 — invert** (the partial-fraction inverse from the previous notebook):

$$
y(t) = -\frac{16}{5}e^{t} + \frac{25}{6}e^{2t} + \frac{1}{30}e^{-4t} .
$$

The initial conditions $y(0)=1$, $y'(0)=5$ were used in Step 1, so no further constants need fixing.

```python
import sympy as sp

t, s = sp.symbols("t s", positive=True)
Y = sp.Symbol("Y")

transformed_equation = (s**2 * Y - s * 1 - 5) - 3 * (s * Y - 1) + 2 * Y - 1 / (s + 4)
Y_solution = sp.solve(transformed_equation, Y)[0]
y_solution = sp.inverse_laplace_transform(Y_solution, s, t)

cross_check = sp.dsolve(sp.Eq(sp.Function("y")(t).diff(t, 2) - 3 * sp.Function("y")(t).diff(t)
                              + 2 * sp.Function("y")(t), sp.exp(-4 * t)), sp.Function("y")(t),
                        ics={sp.Function("y")(0): 1, sp.Function("y")(t).diff(t).subs(t, 0): 5})

print("Y(s) =", sp.simplify(Y_solution))
print("y(t) =", y_solution)
print("dsolve check:", cross_check.rhs)
```
