### 3.2.3.6. Jensen Inequality and Convex Inequalities


$$f(\sum_i\theta_ix_i)\leq\sum_i\theta_if(x_i),\quad\theta_i\geq0,\ \sum_i\theta_i=1$$


**Explanation:**

Jensen inequality extends the two-point convexity inequality to weighted averages. It is the arithmetic form of the chord picture and is often the cleanest way to prove convex inequalities.

The coefficients $\theta_i$ are nonnegative weights that sum to one, so $\sum_i\theta_i x_i$ is a convex combination of points. Jensen inequality says the function value at that average is bounded by the weighted average of the function values.

**Properties:**
- The input average is compared with the average of function values.
- Quadratics with nonnegative curvature are canonical convex examples.
- Jensen inequalities extend the same idea to many points.

**Intuition:**

The computation evaluates both sides of the convexity inequality.


**Numerical Example:**

Let $f(x)=x^2$, $x=0$, $y=2$, and $\theta=1/4$. The averaged input is
$$
\theta x+(1-\theta)y=\frac14\cdot0+\frac34\cdot2=\frac32.
$$

The left side is
$$
f(3/2)=(3/2)^2=9/4.
$$

The right side is
$$
\frac14 f(0)+\frac34 f(2)=\frac14\cdot0+\frac34\cdot4=3.
$$

Since $9/4\leq3$, the convexity inequality holds.


```python
import sympy as sp

theta = sp.Rational(1, 4)
point_x = sp.Integer(0)
point_y = sp.Integer(2)
midpoint = theta * point_x + (1 - theta) * point_y
left_side = midpoint ** 2
right_side = theta * point_x ** 2 + (1 - theta) * point_y ** 2

left_side, right_side, left_side <= right_side
```
