### 1.3.5.12. Implicit Function Theorem

$$
F(x, y) = 0,\ \ \frac{\partial F}{\partial y}(x_0, y_0) \neq 0
\;\Longrightarrow\;
y = y(x)\ \text{locally}, \qquad
\frac{dy}{dx} = -\frac{\partial F/\partial x}{\partial F/\partial y} .
$$

**Explanation:**

The implicit function theorem says that an equation $F(\mathbf{x}, \mathbf{y}) = \mathbf{0}$ can be solved locally for $\mathbf{y}$ as a function of $\mathbf{x}$ whenever the [Jacobian](./07_jacobian.ipynb) block $\partial F/\partial \mathbf{y}$ is invertible at the point, and it gives the derivative of that implicit solution without solving explicitly. The sensitivity formula $\tfrac{dy}{dx} = -F_x/F_y$ is exactly the kind of [parametric sensitivity](../../../03_Optimization/02_Continuous_Optimization/06_sensitivity_of_local_minima.ipynb) used to differentiate the solution of equilibrium and [KKT systems](../../../03_Optimization/03_Convex_Optimization/46_karush_kuhn_tucker_conditions.ipynb) with respect to parameters — the basis of argmin differentiation in modern ML.

**Properties:**
- Vector form: $\dfrac{d\mathbf{y}}{d\mathbf{x}} = -\left(\dfrac{\partial F}{\partial \mathbf{y}}\right)^{-1}\dfrac{\partial F}{\partial \mathbf{x}}$.
- The condition $\partial F/\partial y \neq 0$ fails exactly where the level set has a vertical tangent.

**Numerical Example:**

Let $F(x, y) = x^2 + y^2 - 1 = 0$, the unit circle. Here $F_x = 2x$, $F_y = 2y$, so wherever $y \neq 0$ the circle is locally a function $y(x)$ with

$$
\frac{dy}{dx} = -\frac{F_x}{F_y} = -\frac{2x}{2y} = -\frac{x}{y} .
$$

At the point $\left(\tfrac{1}{\sqrt2}, \tfrac{1}{\sqrt2}\right)$ this gives $\tfrac{dy}{dx} = -1$, the slope of the tangent to the circle there. The theorem fails at $(\pm 1, 0)$, where $F_y = 0$ and the circle has a vertical tangent — no single-valued $y(x)$ exists.

```python
import sympy as sp

x, y = sp.symbols("x y")
F = x**2 + y**2 - 1

partial_x = sp.diff(F, x)
partial_y = sp.diff(F, y)
implicit_slope = -partial_x / partial_y

point = {x: 1 / sp.sqrt(2), y: 1 / sp.sqrt(2)}
slope_at_point = implicit_slope.subs(point)

print("F_x =", partial_x, "  F_y =", partial_y)
print("dy/dx = -F_x/F_y =", implicit_slope)
print("slope at (1/√2, 1/√2) =", sp.simplify(slope_at_point))
print("F_y = 0 (theorem fails) at y =", sp.solve(partial_y, y))
```

**Equivalent CasADi implementation:**

CasADi forms $F_x$ and $F_y$ from the constraint graph and evaluates the implicit-derivative formula at the point.

```python
import casadi as ca

x = ca.SX.sym("x")
y = ca.SX.sym("y")
F = x**2 + y**2 - 1
implicit_slope = -ca.gradient(F, x) / ca.gradient(F, y)
slope_function = ca.Function("slope", [x, y], [implicit_slope])

value = 1 / ca.sqrt(ca.DM(2))
print("dy/dx at (1/√2, 1/√2) =", float(slope_function(value, value)))
```
