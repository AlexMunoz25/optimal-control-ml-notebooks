### 1.1.9.3. Covariant, Contravariant, and Mixed Tensors

$$
A^{\prime ij} = \frac{\partial x^{\prime i}}{\partial x^k}\,\frac{\partial x^{\prime j}}{\partial x^l}\,A^{kl}
\qquad \text{(contravariant rank-2)}
$$

$$
A^{\prime}_{ij} = \frac{\partial x^k}{\partial x^{\prime i}}\,\frac{\partial x^l}{\partial x^{\prime j}}\,A_{kl}
\qquad \text{(covariant rank-2)}
$$

$$
A^{\prime i}{}_{j} = \frac{\partial x^{\prime i}}{\partial x^k}\,\frac{\partial x^l}{\partial x^{\prime j}}\,A^k{}_l
\qquad \text{(mixed rank-2)}
$$

**Explanation:**

- Each upper index transforms with the direct Jacobian.
- Each lower index transforms with the inverse Jacobian.
- A mixed tensor applies the appropriate rule slot by slot.
- This is the rank-2 extension of the contravariant and covariant vector components from the previous module.

**Connections:**

- The [Jacobian Matrix](../08_Coordinate_Transformations/10_jacobian_matrix.ipynb) now appears once for each index.
- Mixed tensors are the natural form of linear maps, because one slot accepts an input component and another returns an output component.

**Example:**

- For the scaling change $x'^1 = 2x^1$ and $x'^2 = \frac{1}{2}x^2$, the Jacobian is
$$
J =
\begin{bmatrix}
2 & 0 \\
0 & \frac{1}{2}
\end{bmatrix}.
$$
- A contravariant tensor transforms as $A' = J A J^\top$.
- A covariant tensor transforms as $A' = J^{-T} A J^{-1}$.
- A mixed tensor transforms as $A' = J A J^{-1}$.


```python
import numpy as np

coordinate_jacobian = np.array([[2.0, 0.0], [0.0, 0.5]])
inverse_coordinate_jacobian = np.linalg.inv(coordinate_jacobian)
base_tensor_components = np.array([[1.0, 2.0], [3.0, 4.0]])

contravariant_tensor_transformed = (
    coordinate_jacobian @ base_tensor_components @ coordinate_jacobian.T
)
covariant_tensor_transformed = (
    inverse_coordinate_jacobian.T @ base_tensor_components @ inverse_coordinate_jacobian
)
mixed_tensor_transformed = (
    coordinate_jacobian @ base_tensor_components @ inverse_coordinate_jacobian
)

print("contravariant_tensor_transformed =\n", contravariant_tensor_transformed)
print("covariant_tensor_transformed =\n", covariant_tensor_transformed)
print("mixed_tensor_transformed =\n", mixed_tensor_transformed)
```
