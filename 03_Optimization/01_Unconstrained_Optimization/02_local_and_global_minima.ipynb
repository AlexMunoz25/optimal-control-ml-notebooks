### 3.1.2. Local and Global Minima


$$
\begin{aligned}
x^* \in \arg\min_{x \in \mathbb{R}^n} f(x)
&\Longleftrightarrow
f(x^*) \le f(x), \quad \forall x \in \mathbb{R}^n, \\
x^* \text{ is local}
&\Longleftrightarrow
\exists \epsilon > 0:\ f(x^*) \le f(x),\quad \forall x \text{ with } \|x-x^*\| < \epsilon.
\end{aligned}
$$


**Explanation:**

In an unconstrained problem, there are no feasibility restrictions other than belonging to the ambient space $\mathbb{R}^n$. This makes the meaning of “minimum” especially important: a global minimizer is compared with every point in the space, while a local minimizer is compared only with points in a sufficiently small neighborhood. Bertsekas begins here because most algorithms in nonlinear programming see only local information, such as nearby function values, gradients, or local quadratic models. A neighborhood is measured by a [vector norm](../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb), so the basic language of local optimality already depends on the geometry introduced in linear algebra.

For a first course, it helps to read the definition as a warning. A point may look best from nearby but still be worse than another point far away, especially when the cost is nonconvex. This is why the later optimality conditions in Chapter 1 are local unless convexity or some other global structure is available.

**Example:**

The figure from Bertsekas shows a one-dimensional cost with several valleys and peaks. It is possible for a point to be locally minimal because every nearby perturbation raises the cost, while another valley gives a smaller value and is therefore the global minimum. The code cell then uses a polynomial example to find stationary points and inspect their second derivatives, illustrating how local candidates arise before any global comparison is made.

<img src="../../Figures/030101_berts_fig_1_1_1_local_global_minima.jpeg" alt="Bertsekas Figure 1.1.1: local and global minima in one dimension" style="width: 60%; height: auto;">


```python
import sympy as sp

variable = sp.symbols("x")
objective = variable**4 - 3 * variable**2 + sp.Rational(7, 20) * variable + sp.Rational(11, 5)
stationary_points = sp.solve(sp.diff(objective, variable), variable)
classified_points = [
    (point, sp.N(objective.subs(variable, point)), sp.N(sp.diff(objective, variable, 2).subs(variable, point)))
    for point in stationary_points
]

classified_points
```
