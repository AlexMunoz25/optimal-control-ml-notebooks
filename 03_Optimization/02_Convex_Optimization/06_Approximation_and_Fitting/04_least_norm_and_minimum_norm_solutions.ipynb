### 3.2.6.4. Least-Norm and Minimum-Norm Solutions


$$\min_x\|x\|\quad\text{subject to}\quad Ax=b$$


**Explanation:**

Least-[norm](../../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb) problems choose the smallest decision vector among all vectors satisfying linear equations. This is useful when the data equation is underdetermined and many feasible solutions exist. Feasibility is exact satisfaction of the equations; optimality selects the feasible point with minimum norm.

The equation $Ax=b$ is a linear constraint, and the objective chooses the feasible vector with smallest [norm](../../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb). When many vectors satisfy the equations, the norm selects one solution from the feasible [affine set](../02_Convex_Sets/01_affine_sets_and_convex_sets.ipynb).

**Properties:**
- The equality constraint defines an affine feasible set.
- The norm objective is convex.
- For Euclidean norm, KKT equations give an exact solution.

**Intuition:**

The example solves a minimum Euclidean norm point on a line.


**Numerical Example:**

Solve
$$
\min_{x_1,x_2}\frac12(x_1^2+x_2^2)
\quad\text{subject to}\quad x_1+x_2=2.
$$

The Lagrangian is
$$
L=\frac12(x_1^2+x_2^2)+\nu(x_1+x_2-2).
$$

Stationarity gives
$$
x_1+\nu=0,
\qquad
x_2+\nu=0.
$$

Thus $x_1=x_2=-\nu$. Substituting into $x_1+x_2=2$ gives
$$
-2\nu=2\quad\Longrightarrow\quad\nu=-1.
$$

Therefore $x_1=x_2=1$.


```python
import sympy as sp

variable_1, variable_2, multiplier = sp.symbols('x_1 x_2 nu')
solution = sp.solve([sp.Eq(variable_1 + multiplier, 0), sp.Eq(variable_2 + multiplier, 0), sp.Eq(variable_1 + variable_2, 2)], [variable_1, variable_2, multiplier])

solution
```
