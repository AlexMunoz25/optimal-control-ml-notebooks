### 1.1.9.11. Christoffel Symbols

$$
\Gamma^k_{ij}\,\mathbf{e}_k = \frac{\partial \mathbf{e}_i}{\partial x^j},
\qquad
\Gamma^l_{ij} = \frac{1}{2}\,g^{kl}\!\left(\frac{\partial g_{ik}}{\partial x^j} + \frac{\partial g_{jk}}{\partial x^i} - \frac{\partial g_{ij}}{\partial x^k}\right)
$$

**Explanation:**

When the basis vectors change from point to point, differentiating a vector field requires differentiating the basis as well as the components. Christoffel symbols are the coefficients in that basis-vector derivative, so they record how each basis direction changes as the coordinates vary. They are built from the metric tensor in [Metric Tensor](./09_metric_tensor.ipynb), and they are the correction terms used in the next notebook on covariant derivatives. They are not tensor components themselves, but for the torsion-free case used here they are symmetric in their lower indices.

**Example:**

For cylindrical coordinates with
$$
g_{ij} = \operatorname{diag}(1, r^2, 1),
$$
the nonzero Christoffel symbols are
$$
\Gamma^r_{\phi\phi} = -r,
\qquad
\Gamma^\phi_{r\phi} = \Gamma^\phi_{\phi r} = \frac{1}{r}.
$$

```python
import sympy as sp

radial_coordinate, angular_coordinate, axial_coordinate = sp.symbols('r phi z', positive=True)
coordinate_list = [radial_coordinate, angular_coordinate, axial_coordinate]
coordinate_labels = ['r', 'phi', 'z']
metric_tensor_covariant = sp.diag(1, radial_coordinate**2, 1)
metric_tensor_contravariant = metric_tensor_covariant.inv()
dimension = len(coordinate_list)


def christoffel_symbol_value(upper_index, first_lower_index, second_lower_index):
    return sp.Rational(1, 2) * sum(
        metric_tensor_contravariant[upper_index, summation_index]
        * (
            sp.diff(
                metric_tensor_covariant[first_lower_index, summation_index],
                coordinate_list[second_lower_index],
            )
            + sp.diff(
                metric_tensor_covariant[second_lower_index, summation_index],
                coordinate_list[first_lower_index],
            )
            - sp.diff(
                metric_tensor_covariant[first_lower_index, second_lower_index],
                coordinate_list[summation_index],
            )
        )
        for summation_index in range(dimension)
    )


index_triplets = [
    (upper_index, first_lower_index, second_lower_index)
    for upper_index in range(dimension)
    for first_lower_index in range(dimension)
    for second_lower_index in range(first_lower_index, dimension)
]
nonzero_symbols = []

for upper_index, first_lower_index, second_lower_index in index_triplets:
    symbol_value = sp.simplify(
        christoffel_symbol_value(upper_index, first_lower_index, second_lower_index)
    )
    if symbol_value != 0:
        nonzero_symbols.append(
            (
                upper_index,
                first_lower_index,
                second_lower_index,
                symbol_value,
            )
        )

for upper_index, first_lower_index, second_lower_index, symbol_value in nonzero_symbols:
    print(
        f"Gamma^{coordinate_labels[upper_index]}_"
        f"{{{coordinate_labels[first_lower_index]}{coordinate_labels[second_lower_index]}}} = {symbol_value}"
    )
```
