### 3.1.3. Necessary Optimality Conditions


$$
\nabla f(x^*) = 0,
\qquad
\nabla^2 f(x^*) \succeq 0.
$$


**Explanation:**

Necessary optimality conditions describe what must be true at a smooth local minimum. If $f$ is differentiable, then moving infinitesimally in any direction cannot produce a first-order decrease, so the gradient must vanish. If $f$ is twice differentiable, the second-order curvature cannot be negative in any direction, which is expressed by positive semidefiniteness of the Hessian in the sense of [matrix definiteness](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/18_matrix_definiteness.ipynb). These conditions are powerful filters, but they are not by themselves proof that a point is a minimum.

The distinction between necessary and sufficient conditions is one of the first conceptual hurdles in optimization. A stationary point can be a local minimum, a local maximum, or a saddle-like point, and a zero second derivative can leave the classification unresolved. Bertsekas uses these examples to keep the reader from treating derivative tests as automatic certificates.

**Example:**

The first figure shows that zero slope and nonnegative curvature can fail to identify a minimum when the function is not convex or when higher-order behavior matters. The second figure gives a more subtle warning: being minimal along every line through a point does not necessarily imply local minimality in the full space. The code cell reinforces the same lesson by comparing simple scalar functions that share stationary behavior but have different local meanings.

```python
import sympy as sp

variable = sp.symbols("x")
objectives = {
    "minimum": variable**2,
    "maximum": -variable**2,
    "flat stationary point": variable**3,
}
second_order_tests = {
    name: (sp.diff(objective, variable).subs(variable, 0), sp.diff(objective, variable, 2).subs(variable, 0))
    for name, objective in objectives.items()
}

second_order_tests
```
