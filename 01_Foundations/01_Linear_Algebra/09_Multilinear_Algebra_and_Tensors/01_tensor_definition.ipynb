### 1.1.9.1. Tensor Definition

$$
\phi' = \phi,
\qquad
\mathbf{A} = A^i \mathbf{e}_i = A_i \mathbf{e}^{\,i},
\qquad
T'^{\,i_1\cdots i_n}
=
\frac{\partial x'^{\,i_1}}{\partial x^{j_1}}
\cdots
\frac{\partial x'^{\,i_n}}{\partial x^{j_n}}
T^{j_1\cdots j_n}
$$

**Explanation:**

The first tensors in this repository already appeared earlier: the scalar discussion in [Vector Definition](../02_Vector/01_vector_definition.ipynb), the vectors in that same notebook, and the matrices in [Matrix Definition](../03_Matrix/01_matrix_definition.ipynb). In tensor language, a scalar is a rank-0 tensor, a vector is a rank-1 tensor, and a matrix is a rank-2 tensor. Higher-rank tensors extend the same idea to more indices, while the defining requirement remains the one emphasized by Fleisch: the components may change with coordinates, but the underlying quantity does not.

**Example:**

If
$$
\phi = 7,
\qquad
v^i =
\begin{bmatrix}
1 \\
2
\end{bmatrix},
\qquad
M^i{}_j =
\begin{bmatrix}
2 & -1 \\
0 & 3
\end{bmatrix},
$$
then $\phi$ is a rank-0 tensor, $v^i$ is a rank-1 tensor, and $M^i{}_j$ is a rank-2 tensor. Using the matrix-vector rule from [Matrix-Vector Multiplication](../03_Matrix/04_matrix_vector_multiplication.ipynb),
$$
w^i = M^i{}_j v^j =
\begin{bmatrix}
0 \\
6
\end{bmatrix},
$$
which is again a rank-1 tensor.


```python
import numpy as np

scalar_tensor = np.array(7.0)
vector_tensor = np.array([1.0, 2.0, 3.0])
matrix_tensor = np.array([[2.0, -1.0], [0.0, 3.0]])
rank_3_tensor = np.arange(8.0).reshape(2, 2, 2)

print("scalar_tensor rank =", scalar_tensor.ndim, "shape =", scalar_tensor.shape)
print("vector_tensor rank =", vector_tensor.ndim, "shape =", vector_tensor.shape)
print("matrix_tensor rank =", matrix_tensor.ndim, "shape =", matrix_tensor.shape)
print("rank_3_tensor rank =", rank_3_tensor.ndim, "shape =", rank_3_tensor.shape)
```
