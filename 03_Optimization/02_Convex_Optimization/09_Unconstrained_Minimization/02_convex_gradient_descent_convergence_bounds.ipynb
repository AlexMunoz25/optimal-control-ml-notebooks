### 3.2.9.2. Convex Gradient Descent Convergence Bounds


$$f(x^{(k)})-p^\star\leq c^k(f(x^{(0)})-p^\star)$$


**Explanation:**

Gradient descent on smooth strongly convex functions admits geometric convergence bounds. The contraction factor depends on curvature constants and the chosen step rule. Feasibility is the whole space; optimality is approached by reducing the objective gap at each iteration.

**Properties:**
- The objective gap contracts by a factor in a standard bound.
- Smaller factors mean faster convergence.
- Curvature bounds determine conservative step-size guarantees.

**Intuition:**

The example applies a contraction bound over two iterations.


**Numerical Example:**

Suppose
$$
f(x^{(0)})-p^\star=16
$$
and the convergence bound uses contraction $c=1/4$.

After one step,
$$
f(x^{(1)})-p^\star\leq\frac14\cdot16=4.
$$

After two steps,
$$
f(x^{(2)})-p^\star\leq\left(\frac14\right)^2 16=1.
$$

The bound predicts geometric decay of the objective gap.


```python
import sympy as sp

initial_gap = sp.Integer(16)
contraction = sp.Rational(1, 4)
gap_after_one_step = contraction * initial_gap
gap_after_two_steps = contraction ** 2 * initial_gap

gap_after_one_step, gap_after_two_steps
```
