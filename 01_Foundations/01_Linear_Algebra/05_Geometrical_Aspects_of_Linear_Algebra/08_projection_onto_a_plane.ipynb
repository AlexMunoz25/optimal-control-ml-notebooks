### 1.1.5.8. Projection onto a Plane


$$
\Pi_{P}(\mathbf{u}) =
\mathbf{u} -
\frac{\mathbf{u} \cdot \mathbf{n}}{\lVert \mathbf{n} \rVert^2} \mathbf{n}
$$


**Explanation:**

The projection onto a plane is obtained by subtracting the normal component of the vector. This is the plane analogue of [projection onto a line](./07_projection_onto_a_line.ipynb), and the result stays inside the plane because its dot product with the normal vector is zero.

**Example:**

Project

$$
\mathbf{u}=\begin{bmatrix}2\\1\\3\end{bmatrix}
$$

onto the plane with normal

$$
\mathbf{n}=\begin{bmatrix}0\\0\\1\end{bmatrix}.
$$

The normal component is

$$
\frac{\mathbf{u}\cdot\mathbf{n}}{\mathbf{n}\cdot\mathbf{n}}\mathbf{n}
=
\frac{2\cdot0+1\cdot0+3\cdot1}{1}\begin{bmatrix}0\\0\\1\end{bmatrix}
=
\begin{bmatrix}0\\0\\3\end{bmatrix}.
$$

Subtracting it gives the projection onto the plane:

$$
\begin{bmatrix}2\\1\\3\end{bmatrix}-\begin{bmatrix}0\\0\\3\end{bmatrix}
=
\begin{bmatrix}2\\1\\0\end{bmatrix}.
$$


```python
import sympy as sp

vector_u = sp.Matrix([2, 1, 3])
plane_normal = sp.Matrix([0, 0, 1])
normal_component = vector_u.dot(plane_normal) / plane_normal.dot(plane_normal) * plane_normal
projection_value = vector_u - normal_component

print("normal_component =", normal_component)
print("projection_value =", projection_value)
```
