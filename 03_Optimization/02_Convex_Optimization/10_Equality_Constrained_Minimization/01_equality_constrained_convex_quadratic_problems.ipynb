### 3.2.10.1. Equality Constrained Convex Quadratic Problems


$$\min_x\frac12x^TPx+q^Tx+r\quad\text{subject to}\quad Ax=b$$


**Explanation:**

Equality constrained convex quadratic problems lead to linear KKT equations. The multiplier enforces the equality constraint while stationarity balances the objective gradient. Feasibility is the equality equation; optimality is stationarity plus feasibility.

**Properties:**
- The Hessian gives convex curvature.
- The equality multiplier enforces the affine constraint.
- The KKT system couples primal variables and multipliers.

**Intuition:**

The example solves a two-variable quadratic with one equality constraint.


**Numerical Example:**

Solve
$$
\min \frac12(x_1^2+x_2^2)
\quad\text{subject to}\quad x_1+x_2=2.
$$

The Lagrangian stationarity equations are
$$
x_1+\nu=0,
\qquad
x_2+\nu=0.
$$

So $x_1=x_2=-\nu$. Substitute into the equality constraint:
$$
-\nu-\nu=2 \quad\Longrightarrow\quad \nu=-1.
$$

Therefore $x_1=x_2=1$.


```python
import sympy as sp

variable_1, variable_2, multiplier = sp.symbols('x_1 x_2 nu')
solution = sp.solve([sp.Eq(variable_1 + multiplier, 0), sp.Eq(variable_2 + multiplier, 0), sp.Eq(variable_1 + variable_2, 2)], [variable_1, variable_2, multiplier])

solution
```
