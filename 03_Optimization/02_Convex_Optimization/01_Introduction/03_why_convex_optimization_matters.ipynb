### 3.2.1.3. Why Convex Optimization Matters


$$\text{convex problem}\quad\Longrightarrow\quad\text{local optimum is global}$$


**Explanation:**

Convexity turns local information into global information. A feasible point that satisfies the right local optimality condition is globally optimal, and dual certificates can bound or certify the best achievable value.

The display is a logical implication. It says that when the objective and feasible set define a convex problem, any feasible point that is locally optimal is already globally optimal.

**Properties:**
- Feasibility is checked before objective values are compared.
- Optimality means no feasible point has a smaller objective value.
- Convex structure makes global certificates possible.

**Intuition:**

Think of the constraints as the search region and the objective as the ruler used inside that region.


**Numerical Example:**

Let $X=\{1,2,3\}$ and $f(x)=(x-2)^2$. First check that all three candidates are feasible because each belongs to $X$.

Then
$$
f(1)=(1-2)^2=1,
$$
$$
f(2)=(2-2)^2=0,
$$
and
$$
f(3)=(3-2)^2=1.
$$

The smallest feasible value is $0$, so the optimizer is $x^\star=2$.


```python
import sympy as sp

candidates = [sp.Integer(1), sp.Integer(2), sp.Integer(3)]
values = [(candidate - 2) ** 2 for candidate in candidates]
optimizer = candidates[values.index(min(values))]

print("optimizer =", optimizer)
print("values =", values)
```
