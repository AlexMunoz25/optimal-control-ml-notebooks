### 3.2.11.11. Interior-Point Implementation Patterns


$$K\Delta z=r$$


**Explanation:**

Interior-point implementations are dominated by structured linear solves. The mathematical residual equations define the right-hand side, and sparse or block structure makes the solve practical at scale.

**Properties:**
- The coefficient matrix represents local curvature and constraints.
- The right-hand side represents residual reduction.
- Structured linear algebra is essential for large conic and barrier problems.

**Intuition:**

The example solves the small linear algebra core exactly.


**Numerical Example:**

Solve
$$
K\Delta z=r,
\qquad
K=\begin{bmatrix}4&1\\1&3\end{bmatrix},
\qquad
r=\begin{bmatrix}1\\2\end{bmatrix}.
$$

The determinant is
$$
4\cdot3-1\cdot1=11.
$$

Therefore
$$
\Delta z_1=\frac{3\cdot1-1\cdot2}{11}=\frac1{11},
\qquad
\Delta z_2=\frac{-1\cdot1+4\cdot2}{11}=\frac7{11}.
$$


```python
import sympy as sp

coefficient_matrix = sp.Matrix([[4, 1], [1, 3]])
right_side = sp.Matrix([1, 2])
solution = coefficient_matrix.LUsolve(right_side)

solution
```
