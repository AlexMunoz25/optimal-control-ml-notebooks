### 3.2.2.2. Affine Hull, Relative Interior, and Dimension


$$\operatorname{aff}C=\{\sum_i\theta_ix_i\mid \sum_i\theta_i=1\}$$


**Explanation:**

The affine hull is the smallest [affine set](01_affine_sets_and_convex_sets.ipynb) containing a set. Relative interior measures interior after restricting attention to that affine hull, and dimension is the dimension of the affine hull. These ideas keep lower-dimensional [convex sets](01_affine_sets_and_convex_sets.ipynb), such as line segments in a plane, from being treated as having empty interior for the wrong reason.

The set $C$ contains points $x_i$, and the coefficients $\theta_i$ are allowed to be signed as long as they sum to one. The display defines the affine hull as all such affine combinations, which are built from [vectors](../../../01_Foundations/01_Linear_Algebra/02_Vector/01_vector_definition.ipynb) in the same ambient space.

**Properties:**
- Affine combinations require coefficients to sum to one.
- Convex combinations also require nonnegative coefficients.
- Every affine set is convex, but a line segment is convex without being affine.

**Intuition:**

The arithmetic computes one specific point on a segment between two endpoints.


**Numerical Example:**

Let
$$
x_1=\begin{bmatrix}0\\0\end{bmatrix},\quad
x_2=\begin{bmatrix}4\\2\end{bmatrix},\quad
\theta=\frac14.
$$

The convex combination is
$$
\theta x_1+(1-\theta)x_2=\frac14\begin{bmatrix}0\\0\end{bmatrix}+\frac34\begin{bmatrix}4\\2\end{bmatrix}.
$$

Coordinate by coordinate,
$$
\frac14\cdot0+\frac34\cdot4=3,
\qquad
\frac14\cdot0+\frac34\cdot2=\frac32.
$$

Therefore the point is
$$
\begin{bmatrix}3\\3/2\end{bmatrix}.
$$


```python
import sympy as sp

point_1 = sp.Matrix([0, 0])
point_2 = sp.Matrix([4, 2])
theta = sp.Rational(1, 4)
combination = theta * point_1 + (1 - theta) * point_2

print("combination =", combination)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

point_1 = ca.DM([0, 0])
point_2 = ca.DM([4, 2])
theta = 0.25
combination = theta * point_1 + (1 - theta) * point_2

print("combination =", combination)
```
