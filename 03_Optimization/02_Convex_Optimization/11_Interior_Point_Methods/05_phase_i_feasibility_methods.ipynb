### 3.2.11.5. Phase I Feasibility Methods


$$\min_{x,s}s\quad\text{subject to}\quad f_i(x)\leq s$$


**Explanation:**

Phase I methods solve an auxiliary optimization problem whose goal is to find a feasible starting point. A slack variable measures the worst constraint violation. Feasibility of the original inequalities is recovered when the optimal slack is nonpositive.

**Properties:**
- Positive slack means some constraint is violated.
- Nonpositive slack gives original feasibility.
- Phase I prepares an initial point for Phase II barrier methods.

**Intuition:**

The example computes the minimum slack needed for one violated inequality.


**Numerical Example:**

Suppose the original inequality is
$$
x-2\leq0
$$
and the trial point is $x=3$. The violation is
$$
3-2=1.
$$

Phase I introduces $s$ and requires
$$
x-2\leq s.
$$

At $x=3$, this becomes
$$
1\leq s.
$$

The smallest feasible slack is $s=1$.


```python
import sympy as sp

point = sp.Integer(3)
constraint_violation = point - 2
minimum_slack = constraint_violation

minimum_slack
```
