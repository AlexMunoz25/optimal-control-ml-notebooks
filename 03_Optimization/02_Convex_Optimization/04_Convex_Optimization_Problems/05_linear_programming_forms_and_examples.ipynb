### 3.2.4.5. Linear Programming Forms and Examples


$$\min_x c^Tx\quad\text{subject to}\quad Ax\preceq b$$


**Explanation:**

Linear programs minimize a linear objective over a polyhedron. Feasibility is described by affine inequalities, and optimality can be understood geometrically by sliding objective hyperplanes until they first touch the feasible set.

The vector $x$ is the decision variable and $c^Tx$ is a linear objective formed by an [inner product](../../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/02_inner_product_space.ipynb). The matrix inequality $Ax\preceq b$ represents a finite set of affine inequality constraints.

**Properties:**
- The feasible set is an intersection of halfspaces.
- The objective has affine level sets.
- An optimum, when finite and attained, occurs on a face of the polyhedron.

**Intuition:**

The figure above shows linear objective geometry over a feasible polygon; the calculation compares feasible vertices.

**Numerical Example:**

The figure above shows the same vertex-comparison geometry for a linear program.

Let
$$
c=\begin{bmatrix}1\\2\end{bmatrix}
$$
and compare feasible vertices
$$
x^{(1)}=\begin{bmatrix}1\\0\end{bmatrix},\quad
x^{(2)}=\begin{bmatrix}0\\1\end{bmatrix},\quad
x^{(3)}=\begin{bmatrix}1\\1\end{bmatrix}.
$$

The objective values are
$$
c^Tx^{(1)}=1,
\qquad
c^Tx^{(2)}=2,
\qquad
c^Tx^{(3)}=3.
$$

Among these feasible candidates, the smallest value is $1$ at $x^{(1)}$.


```python
import sympy as sp

cost_vector = sp.Matrix([1, 2])
candidate_points = [sp.Matrix([1, 0]), sp.Matrix([0, 1]), sp.Matrix([1, 1])]
costs = [(cost_vector.T * candidate)[0] for candidate in candidate_points]
optimizer = candidate_points[costs.index(min(costs))]

print("optimizer =", optimizer)
print("costs =", costs)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

cost_vector = ca.DM([1, 2])
candidate_points = [ca.DM([1, 0]), ca.DM([0, 1]), ca.DM([1, 1])]
costs = [float(ca.dot(cost_vector, candidate)) for candidate in candidate_points]
optimizer = candidate_points[costs.index(min(costs))]

print("optimizer =", optimizer)
print("costs =", costs)
```
