### 3.1.9. Stepsize Rules


$$
\begin{aligned}
\alpha^k &\in \arg\min_{\alpha \ge 0} f(x^k+\alpha d^k), \\
f(x^k)-f(x^k+\beta^m s d^k)
&\ge -\sigma \beta^m s\nabla f(x^k)^{\mathsf T}d^k, \\
\sigma
&\le
\frac{f(x^k+\alpha^k d^k)-f(x^k)}{\alpha^k\nabla f(x^k)^{\mathsf T}d^k}
\le 1-\sigma.
\end{aligned}
$$


**Explanation:**

Once a descent direction has been chosen, the stepsize determines whether the algorithm makes useful progress. Exact line minimization is conceptually attractive because it chooses the best point along the ray, but it can be too expensive or unnecessary in practice. Bertsekas therefore introduces rules that require enough decrease rather than exact minimization, especially the Armijo rule and the Goldstein rule. The common thread is the directional derivative $\nabla f(x^k)^{\mathsf T}d^k$, written with [matrix transpose](../../01_Foundations/01_Linear_Algebra/03_Matrix/06_matrix_transpose.ipynb) notation, which predicts the first-order decrease available along the chosen direction.

A beginner-friendly way to view these rules is as safeguards against two bad extremes. A step that is too large may overshoot and increase the cost; a step that is technically decreasing but too small may waste iterations and fail to drive the gradient toward zero. Sufficient decrease rules give a practical compromise between reliability and computational cost.

**Example:**

The first figure shows why accepting any cost-reducing step can still fail. Armijo backtracking fixes this by requiring the observed decrease to be large enough relative to the predicted linear decrease. Goldstein's rule places the step in an acceptable interval that avoids both overly short and overly long steps. The code cell implements the Armijo idea for a simple quadratic direction.

```python
import numpy as np

Q = np.array([[5.0, 0.0], [0.0, 1.0]])
current_point = np.array([1.0, 2.0])
gradient = Q @ current_point
direction = -gradient
trial_stepsize = 1.0
reduction_factor = 0.5
sufficient_decrease = 1e-4
objective = lambda point: 0.5 * point @ Q @ point

while objective(current_point) - objective(current_point + trial_stepsize * direction) < -sufficient_decrease * trial_stepsize * gradient @ direction:
    trial_stepsize *= reduction_factor

trial_stepsize, objective(current_point + trial_stepsize * direction)
```
