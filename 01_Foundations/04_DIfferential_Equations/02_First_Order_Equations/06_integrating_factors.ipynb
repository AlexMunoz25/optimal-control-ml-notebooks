### 1.4.2.6. Integrating Factors

$$
\frac{M_y - N_x}{N} = \text{function of } x \ \Rightarrow\ \mu(x) = e^{\int \frac{M_y - N_x}{N}\,dx};
\qquad
\frac{N_x - M_y}{M} = \text{function of } y \ \Rightarrow\ \mu(y) = e^{\int \frac{N_x - M_y}{M}\,dy} .
$$

**Explanation:**

When $M\,dx + N\,dy = 0$ is **not** [exact](./05_exact_equations.ipynb), it can sometimes be made exact by multiplying through by an **integrating factor** $\mu$. Finding a general $\mu(x,y)$ requires a PDE, but two special cases are easy: if $(M_y - N_x)/N$ depends on $x$ alone, there is a factor $\mu(x)$; if $(N_x - M_y)/M$ depends on $y$ alone, there is a factor $\mu(y)$. Each $\mu$ is itself the solution of a simple separable/linear equation. The [linear first-order](./04_linear_first_order_equations.ipynb) integrating factor $e^{\int P\,dx}$ is exactly this construction applied to $(Py - f)\,dx + dy = 0$, where $(M_y - N_x)/N = P(x)$.

**Properties:**
- After multiplying by $\mu$, re-test $\partial(\mu M)/\partial y = \partial(\mu N)/\partial x$ before solving.
- The choice between $\mu(x)$ and $\mu(y)$ is decided by which test quotient depends on a single variable.

**Numerical Example:**

Solve $xy\,dx + (2x^2 + 3y^2 - 20)\,dy = 0$.

**Step 1 — check exactness.** $M = xy$, $N = 2x^2 + 3y^2 - 20$, so $M_y = x$ and $N_x = 4x$; since $M_y \neq N_x$ the equation is **not exact**.

**Step 2 — search for a factor.** The $x$-quotient

$$
\frac{M_y - N_x}{N} = \frac{x - 4x}{2x^2 + 3y^2 - 20} = \frac{-3x}{2x^2 + 3y^2 - 20}
$$

depends on both variables — no $\mu(x)$. The $y$-quotient, however,

$$
\frac{N_x - M_y}{M} = \frac{4x - x}{xy} = \frac{3x}{xy} = \frac{3}{y}
$$

depends on $y$ alone, so

$$
\mu(y) = e^{\int (3/y)\,dy} = e^{3\ln y} = y^{3}.
$$

**Step 3 — multiply and solve.** Multiplying by $y^3$ gives the now-exact equation

$$
xy^4\,dx + (2x^2 y^3 + 3y^5 - 20y^3)\,dy = 0,
$$

whose potential (by the exact-equation method) yields the family

$$
\tfrac{1}{2}x^2 y^4 + \tfrac{1}{2}y^6 - 5y^4 = c .
$$

```python
import sympy as sp

x, y = sp.symbols("x y")
M = x * y
N = 2 * x**2 + 3 * y**2 - 20

quotient_x = sp.simplify((sp.diff(M, y) - sp.diff(N, x)) / N)
quotient_y = sp.simplify((sp.diff(N, x) - sp.diff(M, y)) / M)
integrating_factor = sp.exp(sp.integrate(quotient_y, y))

multiplied_M = sp.expand(integrating_factor * M)
multiplied_N = sp.expand(integrating_factor * N)
now_exact = sp.simplify(sp.diff(multiplied_M, y) - sp.diff(multiplied_N, x)) == 0

print("(M_y - N_x)/N =", quotient_x, " -> depends on x and y")
print("(N_x - M_y)/M =", quotient_y, " -> function of y alone")
print("integrating factor mu(y) =", integrating_factor)
print("after multiplying, exact :", now_exact)
```
