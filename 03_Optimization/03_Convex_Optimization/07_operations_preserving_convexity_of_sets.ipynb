### 3.3.7. Operations Preserving Convexity of Sets

$$C_1\cap C_2\ \text{is convex if}\ C_1,C_2\ \text{are convex}$$


**Explanation:**

[Convex sets](01_affine_sets_and_convex_sets.ipynb) can be built from simpler convex sets without returning to the definition each time. Intersections, affine images, affine preimages, and several standard transformations preserve convexity, which makes them basic modeling tools for feasible regions.

The display says that the intersection keeps only points that belong to both $C_1$ and $C_2$. If each set contains every line segment between its own points, then their common points still contain every segment between common points.

**Properties:**
- Affine combinations require coefficients to sum to one.
- Convex combinations also require nonnegative coefficients.
- Every [affine set](01_affine_sets_and_convex_sets.ipynb) is convex, but a line segment is convex without being affine.

**Intuition:**

The arithmetic computes one specific point on a segment between two endpoints.


**Numerical Example:**

Let
$$
x_1=\begin{bmatrix}0\\0\end{bmatrix},\quad
x_2=\begin{bmatrix}4\\2\end{bmatrix},\quad
\theta=\frac14.
$$

The convex combination is
$$
\theta x_1+(1-\theta)x_2=\frac14\begin{bmatrix}0\\0\end{bmatrix}+\frac34\begin{bmatrix}4\\2\end{bmatrix}.
$$

Coordinate by coordinate,
$$
\frac14\cdot0+\frac34\cdot4=3,
\qquad
\frac14\cdot0+\frac34\cdot2=\frac32.
$$

Therefore the point is
$$
\begin{bmatrix}3\\3/2\end{bmatrix}.
$$


```python
import sympy as sp

point_1 = sp.Matrix([0, 0])
point_2 = sp.Matrix([4, 2])
theta = sp.Rational(1, 4)
combination = theta * point_1 + (1 - theta) * point_2

print("combination =", combination)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

point_1 = ca.DM([0, 0])
point_2 = ca.DM([4, 2])
theta = 0.25
combination = theta * point_1 + (1 - theta) * point_2

print("combination =", combination)
```
