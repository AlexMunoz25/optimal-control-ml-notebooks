**Explanation:**

The Cartesian plane is the two-dimensional generalization of the number line.
The horizontal axis is called the $x$-axis, and the vertical axis is the $y$-axis.
Every point in the plane corresponds to a pair of real numbers $(x, y)$.

Because points in the plane are ordered pairs of real numbers, the notation from [Sets and Set Notation](./01_sets.ipynb) remains active here. The same geometric picture becomes the language of [Vectors](../02_Vector/01_vector_definition.ipynb) and is later contrasted with [Non-Cartesian Coordinate Systems](./03_non_cartesian_coordinate_systems.ipynb).

**Example:**

If $P = (2, -1)$, the point lies two units to the right of the origin and one unit below the $x$-axis.

For a function $f(x) = x^2$, each $x$-value maps to a corresponding $y$-value forming a parabola in the plane.


```python
import numpy as np
import matplotlib.pyplot as plt

x_values = np.linspace(-3, 3, 200)
y_values = x_values ** 2
point = (2, -1)
sample_x = 2
sample_y = sample_x ** 2

print("point =", point)
print("f(2) =", sample_y)

plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, linewidth=2, label="$f(x) = x^2$")
plt.scatter([point[0]], [point[1]], color="red", s=100, zorder=5, label="$P = (2, -1)$")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.xlabel("x")
plt.ylabel("y")
plt.title("The Cartesian Plane")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
