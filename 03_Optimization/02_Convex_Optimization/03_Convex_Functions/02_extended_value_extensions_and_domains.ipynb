### 3.2.3.2. Extended-Value Extensions and Domains


$$\tilde f(x)=f(x)\ \text{on dom }f,\qquad \tilde f(x)=\infty\ \text{outside dom }f$$


**Explanation:**

Extended-value notation folds domain restrictions into the objective by assigning +infinity outside the allowed set. This is a compact way to treat constraints and objectives together while preserving convexity when the domain is convex.

The extended function $\tilde f$ agrees with $f$ on its domain and takes value $\infty$ outside that domain. This convention lets a domain restriction be read as part of the objective value.

**Properties:**
- The log barrier requires every inequality value to be negative.
- The barrier is finite in the strict interior.
- Interior-point methods maintain strict feasibility while reducing the approximation gap.

**Intuition:**

The example evaluates one barrier contribution at a strictly feasible point.


**Numerical Example:**

For the constraint $f_1(x)=x-2<0$, evaluate the point $x=1$. Then
$$
f_1(1)=1-2=-1.
$$

The point is strictly feasible because $-1<0$. The log-barrier contribution is
$$
-\log(-f_1(1))=-\log(-(-1))=-\log(1)=0.
$$

The finite value confirms that the point is in the strict interior.


```python
import sympy as sp

point = sp.Integer(1)
constraint_value = point - 2
barrier_value = -sp.log(-constraint_value)

print("barrier_value =", barrier_value)
```
