### 1.3.5.3. Gradient

$$
\nabla f(\mathbf{x}) =
\begin{bmatrix}
\dfrac{\partial f}{\partial x_1} & \dfrac{\partial f}{\partial x_2} & \cdots & \dfrac{\partial f}{\partial x_n}
\end{bmatrix}^{\!\top}.
$$

**Explanation:**

The gradient packages all first-order [partial derivatives](./02_partial_derivatives.ipynb) of a scalar field into a single vector. It points in the direction of steepest *increase* of $f$, its negative points in the direction of steepest *decrease*, and its magnitude is the maximal rate of change at that point. This is precisely what gradient descent follows: each step moves along $-\nabla f$ to reduce a [cost function](../../../03_Optimization/01_Unconstrained_Optimization/08_gradient_method_framework.ipynb). The gradient also pairs with a direction through the [inner product](../../01_Linear_Algebra/02_Vector/04_vector_dot_product.ipynb) to give the [directional derivative](./04_directional_derivative.ipynb).

**Properties:**
- $\nabla f = \mathbf{0}$ at every interior extremum (the multivariable first-order necessary condition).
- The gradient is the transpose of the $1 \times n$ Jacobian of a scalar-valued function.

**Numerical Example:**

For $f(x, y) = x^2 + y^2$, the partials are $f_x = 2x$ and $f_y = 2y$, so

$$
\nabla f(x, y) = \begin{bmatrix} 2x \\ 2y \end{bmatrix}, \qquad
\nabla f(1, 2) = \begin{bmatrix} 2 \\ 4 \end{bmatrix}.
$$

The steepest-increase direction at $(1,2)$ is the unit gradient $\tfrac{1}{\sqrt{20}}[2, 4]^\top$, and the maximal rate of increase there is $\|\nabla f(1,2)\| = \sqrt{2^2 + 4^2} = \sqrt{20} \approx 4.47$.

```python
import sympy as sp

x, y = sp.symbols("x y")
f = x**2 + y**2
variables = sp.Matrix([x, y])

gradient = sp.Matrix([sp.diff(f, variable) for variable in variables])
gradient_at_point = gradient.subs({x: 1, y: 2})
steepest_rate = gradient_at_point.norm()

print("∇f(x,y) ="); sp.pprint(gradient)
print("∇f(1,2) =", gradient_at_point.T, " max rate ‖∇f‖ =", sp.N(steepest_rate, 6))
```

**Equivalent CasADi implementation:**

`ca.gradient` returns the same gradient vector from the computational graph.

```python
import casadi as ca

variables = ca.SX.sym("x", 2)
f = variables[0]**2 + variables[1]**2
gradient = ca.gradient(f, variables)
gradient_function = ca.Function("grad", [variables], [gradient])

print("∇f(1,2) =", gradient_function([1, 2]))
```
