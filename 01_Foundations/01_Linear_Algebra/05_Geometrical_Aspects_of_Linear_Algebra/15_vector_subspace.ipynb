### 1.1.5.15. Vector Subspace


$$
W \subseteq V,
\qquad
\mathbf{w}_1, \mathbf{w}_2 \in W,
\qquad
\alpha \in \mathbb{R}
\Longrightarrow
\alpha \mathbf{w}_1 + \mathbf{w}_2 \in W
$$


**Explanation:**

A subspace is a subset of a [vector space](./13_vector_space.ipynb) that is itself a vector space under the same operations. Savov contrasts valid subspaces with ordinary subsets to show why closure matters. This notebook is kept separate from [span](./14_span.ipynb) because the concept here is the closure test, not the generating set.

**Example:**

The set

$$
W = \left\{
\begin{bmatrix}
x \\
y \\
0
\end{bmatrix}
\mid x, y \in \mathbb{R}
\right\}
$$

is a subspace of $\mathbb{R}^3$ because sums and scalar multiples stay in the plane $z = 0$.


```python
import sympy as sp

vector_w1 = sp.Matrix([1, 2, 0])
vector_w2 = sp.Matrix([3, -1, 0])
scalar_value = 2
closure_test = scalar_value * vector_w1 + vector_w2

print("closure_test =", closure_test)
print("third_coordinate =", closure_test[2])
```
