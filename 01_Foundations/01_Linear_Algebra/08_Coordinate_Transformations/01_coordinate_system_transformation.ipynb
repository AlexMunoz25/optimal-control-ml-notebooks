### 1.1.8.1. Coordinate-System Transformation


$$
\begin{aligned}
A^{x'} &= A^x \cos(\alpha_{11}) + A^y \cos(\alpha_{12}), \\
A^{y'} &= A^x \cos(\alpha_{21}) + A^y \cos(\alpha_{22}), \\
\begin{bmatrix}
A^{x'} \\
A^{y'}
\end{bmatrix}
&=
\begin{bmatrix}
\cos(\alpha_{11}) & \cos(\alpha_{12}) \\
\cos(\alpha_{21}) & \cos(\alpha_{22})
\end{bmatrix}
\begin{bmatrix}
A^x \\
A^y
\end{bmatrix}
=
\begin{bmatrix}
\cos \theta & \sin \theta \\
-\sin \theta & \cos \theta
\end{bmatrix}
\begin{bmatrix}
A^x \\
A^y
\end{bmatrix}
\end{aligned}
$$


**Explanation:**

Rotating the coordinate axes does not change the geometric vector; it only changes the numerical components used to describe that vector. Because the new axes point partly along the old directions, each new component is a weighted linear combination of the original components. The cosine factors are the direction cosines between the original and rotated axes.

**Example:**

For $\theta=150^\circ=5\pi/6$,

$$
\cos\theta=-\frac{\sqrt3}{2},
\qquad
\sin\theta=\frac12.
$$

The passive transformation matrix is

$$
P=\begin{bmatrix}-\sqrt3/2&1/2\\-1/2&-\sqrt3/2\end{bmatrix}.
$$

For components $\begin{bmatrix}5\\3\end{bmatrix}$,

$$
P\begin{bmatrix}5\\3\end{bmatrix}
=
\begin{bmatrix}(-\sqrt3/2)5+(1/2)3\\(-1/2)5+(-\sqrt3/2)3\end{bmatrix}
=
\begin{bmatrix}3/2-5\sqrt3/2\\-5/2-3\sqrt3/2\end{bmatrix}.
$$


```python
import sympy as sp

rotation_angle = sp.rad(150)
passive_transformation_matrix = sp.Matrix([
    [sp.cos(rotation_angle), sp.sin(rotation_angle)],
    [-sp.sin(rotation_angle), sp.cos(rotation_angle)],
])
original_vector_components = sp.Matrix([5, 3])
transformed_vector_components = passive_transformation_matrix * original_vector_components

print("passive_transformation_matrix =")
sp.pprint(passive_transformation_matrix)
print("transformed_vector_components =")
sp.pprint(transformed_vector_components)
```
