### 1.3.1.5. Asymptotic Notation (Big-O, Little-o)

$$
f = O(g) \ \text{as } x \to a \iff \limsup_{x\to a}\left|\frac{f(x)}{g(x)}\right| < \infty,
\qquad
f = o(g) \ \text{as } x \to a \iff \lim_{x\to a}\frac{f(x)}{g(x)} = 0 .
$$

**Explanation:**

Asymptotic notation compares how two functions behave near a point $a$ (often $a = 0$) or as $x \to \infty$, using the limit of their ratio. Big-O, $f = O(g)$, means $|f|$ is bounded by a constant multiple of $|g|$ in the limit ŌĆö $f$ grows no faster than $g$. Little-o, $f = o(g)$, is the stronger statement that $f$ becomes *negligible* relative to $g$, since their ratio tends to $0$. This is the exact vocabulary used to state the error term of a [Taylor approximation](../04_Sequences_Series_and_Taylor_Approximation/05_taylors_theorem_with_remainder.ipynb) and the convergence rates of optimization methods.

**Properties:**
- $o(g) \subset O(g)$: every little-o relation is also big-O, but not conversely.
- As $x\to\infty$, polynomials are ordered by degree: $x^{m} = o(x^{n})$ whenever $m < n$.
- Constants and lower-order terms are absorbed: $2x^2 + 3x = O(x^2)$ and $2x^2 + 3x = x^2\,(2 + o(1))$.

**Numerical Example:**

As $x \to \infty$, compare $x^2$ and $x^3$:

$$
\lim_{x\to\infty} \frac{x^2}{x^3} = \lim_{x\to\infty}\frac{1}{x} = 0 \;\Rightarrow\; x^2 = o(x^3),
$$

so $x^2$ is asymptotically negligible against $x^3$.

As $x \to 0$, the error of the linear approximation $\sin x \approx x$ is

$$
\sin x - x = -\frac{x^3}{6} + \cdots \;\Rightarrow\; \lim_{x\to 0}\frac{\sin x - x}{x^2} = 0,
$$

hence $\sin x - x = o(x^2)$ ŌĆö the remainder vanishes faster than $x^2$. The series form makes the order explicit: $\sin x = x - \tfrac{x^3}{6} + O(x^5)$.

```python
import sympy as sp

x = sp.symbols("x", positive=True)

growth_ratio = sp.limit(x**2 / x**3, x, sp.oo)
big_o_ratio = sp.limit((2 * x**2 + 3 * x) / x**2, x, sp.oo)

remainder = sp.sin(x) - x
little_o_ratio = sp.limit(remainder / x**2, x, 0)
sine_series = sp.series(sp.sin(x), x, 0, 6)

print("lim x->oo  x^2/x^3 =", growth_ratio, " => x^2 = o(x^3)")
print("lim x->oo  (2x^2+3x)/x^2 =", big_o_ratio, " => 2x^2+3x = O(x^2)")
print("lim x->0   (sin x - x)/x^2 =", little_o_ratio, " => sin x - x = o(x^2)")
print("series sin(x) =", sine_series)
```
