### 1.1.8.6. Dual Basis


$$
\mathbf{e}^{\,i} \cdot \mathbf{e}_j = \delta^i_j,
\qquad
\mathbf{A} = A_i \mathbf{e}^{\,i},
\qquad
\mathbf{A} = A^i \mathbf{e}_i = A_i \mathbf{e}^{\,i}
$$


**Explanation:**

The dual basis is defined so that each dual vector is orthogonal to every primal basis vector with a different index and has unit dot product with its matching basis vector. This biorthogonality makes perpendicular projections into valid components, because covariant components combine naturally with dual basis vectors. In an orthonormal basis, the primal and dual bases coincide, so the distinction disappears.

  <img src="../../../Figures/01010803_covariant_components_dual_basis.jpeg" alt="Dual basis vectors in a non-orthogonal system" width="380">
</p>

**Example:**

For
$$
\mathbf{e}_1 =
\begin{bmatrix}
1 \\
3
\end{bmatrix},
\qquad
\mathbf{e}_2 =
\begin{bmatrix}
4 \\
0
\end{bmatrix},
$$
the dual basis is
$$
\mathbf{e}^{\,1} =
\begin{bmatrix}
0 \\
\frac{1}{3}
\end{bmatrix},
\qquad
\mathbf{e}^{\,2} =
\begin{bmatrix}
\frac{1}{4} \\
-\frac{1}{12}
\end{bmatrix},
$$
which satisfies $\mathbf{e}^{\,i} \cdot \mathbf{e}_j = \delta^i_j$.


```python
import sympy as sp

basis_vector_1 = sp.Matrix([1, 3])
basis_vector_2 = sp.Matrix([4, 0])
primal_basis_matrix = sp.Matrix.hstack(basis_vector_1, basis_vector_2)
dual_basis_matrix = primal_basis_matrix.inv().T

print("dual_basis_matrix =")
sp.pprint(dual_basis_matrix)
print("biorthogonality_check =")
sp.pprint(sp.simplify(dual_basis_matrix.T * primal_basis_matrix))
```
