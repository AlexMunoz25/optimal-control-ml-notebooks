### 3.3.4. Hyperplanes, Halfspaces, Balls, and Ellipsoids

$$\{x\mid a^Tx=b\},\qquad\{x\mid a^Tx\leq b\}$$


**Explanation:**

Hyperplanes, halfspaces, balls, and ellipsoids give the most common geometric building blocks for convex feasible regions. Hyperplanes and halfspaces come from affine inequalities, while balls and ellipsoids express [distance](../../01_Foundations/01_Linear_Algebra/05_Geometrical_Aspects_of_Linear_Algebra/04_distance_between_points.ipynb) or quadratic-size limits.

Here $x$ is the point being tested, $a$ is a normal vector, and $b$ is a scalar offset. The expression $a^Tx$ is an [inner product](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/02_inner_product_space.ipynb); equality gives a hyperplane, while the inequality chooses one side of that hyperplane.

**Properties:**
- A hyperplane satisfies one affine equality.
- A halfspace satisfies one affine inequality.
- Finite intersections of halfspaces and hyperplanes form polyhedra.

**Intuition:**

The figure above emphasizes the normal-vector side test; the arithmetic applies that test to one point.

<img src="../../Figures/030204_bv_fig_2_6_hyperplane_normal.jpeg" alt="Boyd and Vandenberghe Figure 2.6: hyperplane and normal vector" style="width: 60%; height: auto;">


**Numerical Example:**

The figure above emphasizes the normal-vector side test; the arithmetic applies that test to one point.

Let
$$
a=\begin{bmatrix}1\\2\end{bmatrix},\quad
b=5,\quad
x=\begin{bmatrix}1\\2\end{bmatrix}.
$$

Compute the halfspace test:
$$
a^Tx=1\cdot1+2\cdot2=1+4=5.
$$

Since
$$
5\leq5,
$$
$x$ is feasible. Equality holds, so $x$ lies on the boundary hyperplane.


```python
import sympy as sp

normal_vector = sp.Matrix([1, 2])
point = sp.Matrix([1, 2])
boundary_value = sp.Integer(5)
inner_product = (normal_vector.T * point)[0]

print("inner_product =", inner_product)
print("inner_product <= boundary_value =", inner_product <= boundary_value)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

normal_vector = ca.DM([1, 2])
point = ca.DM([1, 2])
boundary_value = 5
inner_product = ca.dot(normal_vector, point)

print("inner_product =", inner_product)
print("inner_product <= boundary_value =", bool(inner_product <= boundary_value))
```
