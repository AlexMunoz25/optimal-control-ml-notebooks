### 3.2.9.3. Steepest Descent in General Norms


$$\Delta x_{\rm sd}=\arg\min_v\nabla f(x)^Tv\quad\text{subject to}\quad\|v\|\leq1$$


**Explanation:**

Steepest descent depends on the norm used to measure the step. The direction minimizes the linearized objective over the unit ball of that norm. Feasibility is the step-size unit ball; optimality is the most negative directional derivative.

The vector $v$ ranges over the unit ball of the chosen [norm](../../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb). The steepest descent direction $\Delta x_{\rm sd}$ is the feasible vector that makes the linearized objective $\nabla f(x)^Tv$ as small as possible.

**Properties:**
- Different [norms](../../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb) can produce different steepest directions.
- The Euclidean direction is the negative normalized gradient.
- The dual norm controls the maximum linear decrease.

**Intuition:**

The example compares Euclidean and infinity-norm steepest directions.


**Numerical Example:**

Let
$$
g=\begin{bmatrix}2\\-1\end{bmatrix}.
$$

The Euclidean norm is
$$
\|g\|_2=\sqrt{2^2+(-1)^2}=\sqrt5.
$$

The Euclidean steepest descent direction is
$$
-\frac{g}{\|g\|_2}=\begin{bmatrix}-2/\sqrt5\\1/\sqrt5\end{bmatrix}.
$$

Over the infinity-norm unit ball, the componentwise sign direction is $[-1,1]^T$.


```python
import sympy as sp

gradient = sp.Matrix([2, -1])
euclidean_norm = sp.sqrt((gradient.T * gradient)[0])
euclidean_direction = -gradient / euclidean_norm
infinity_norm_direction = sp.Matrix([-1, 1])

print("euclidean_direction =", euclidean_direction)
print("infinity_norm_direction =", infinity_norm_direction)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

gradient = ca.DM([2, -1])
euclidean_norm = ca.norm_2(gradient)
euclidean_direction = -gradient / euclidean_norm
infinity_norm_direction = ca.DM([-1, 1])

print("euclidean_direction =", euclidean_direction)
print("infinity_norm_direction =", infinity_norm_direction)
```
