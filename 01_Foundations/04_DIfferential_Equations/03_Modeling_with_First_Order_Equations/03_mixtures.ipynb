### 1.4.3.3. Mixtures

$$
\frac{dA}{dt} = R_{\text{in}} - R_{\text{out}},
\qquad R_{\text{in}} = c_{\text{in}}\,r_{\text{in}}, \qquad
R_{\text{out}} = \frac{A}{V}\,r_{\text{out}} .
$$

**Explanation:**

When brine flows into and out of a stirred tank, the amount of salt $A(t)$ changes at the *net* rate (input rate of salt) minus (output rate of salt). The input rate is the inflow concentration times inflow speed; the output rate is the *current tank concentration* $A/V$ times the outflow speed. When inflow and outflow speeds are equal the volume $V$ is constant and the model is a [linear](../02_First_Order_Equations/04_linear_first_order_equations.ipynb) first-order equation. Mixing models are the elementary version of the mass-balance compartment models used throughout process and pharmacokinetic modeling.

**Numerical Example:**

A tank holds 300 gal of brine. Brine of concentration $2$ lb/gal enters at $3$ gal/min and the stirred mixture leaves at $3$ gal/min; initially $50$ lb of salt is dissolved.

Input rate of salt: $R_{\text{in}} = (2\ \text{lb/gal})(3\ \text{gal/min}) = 6$ lb/min.
Output rate of salt: the concentration is $A/300$ lb/gal, so $R_{\text{out}} = \dfrac{A}{300}(3) = \dfrac{A}{100}$ lb/min. Hence

$$
\frac{dA}{dt} = 6 - \frac{A}{100} \;\Longrightarrow\; \frac{dA}{dt} + \frac{1}{100}A = 6, \quad A(0) = 50 .
$$

This is linear with integrating factor $\mu = e^{t/100}$:

$$
\frac{d}{dt}\big[e^{t/100}A\big] = 6e^{t/100} \;\Rightarrow\; e^{t/100}A = 600\,e^{t/100} + c \;\Rightarrow\; A = 600 + c e^{-t/100}.
$$

Apply $A(0) = 50$: $\;50 = 600 + c \Rightarrow c = -550$. Therefore

$$
A(t) = 600 - 550\,e^{-t/100},
$$

and $A(t) \to 600$ lb as $t \to \infty$ — exactly $(300\ \text{gal})(2\ \text{lb/gal})$, the inflow concentration filling the whole tank.

```python
import sympy as sp

t = sp.symbols("t")
A = sp.Function("A")

mixing = sp.Eq(A(t).diff(t) + A(t) / 100, 6)
solution = sp.dsolve(mixing, A(t), ics={A(0): 50})
steady_state = sp.limit(solution.rhs, t, sp.oo)

print("dA/dt + A/100 = 6,  A(0) = 50")
print("A(t) =", solution.rhs)
print("long-run salt =", steady_state, "lb")
```
