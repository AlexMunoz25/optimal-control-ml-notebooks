### 1.3.7.3. Gradients of Linear and Quadratic Forms

$$
\nabla_{\mathbf{x}}\big(\mathbf{a}^{\top}\mathbf{x}\big) = \mathbf{a},
\qquad
\nabla_{\mathbf{x}}\big(\mathbf{x}^{\top} Q \mathbf{x}\big) = (Q + Q^{\top})\mathbf{x}
\;\xrightarrow{\,Q = Q^{\top}\,}\; 2Q\mathbf{x} .
$$

**Explanation:**

The linear form $\mathbf{a}^\top\mathbf{x}$ has the constant gradient $\mathbf{a}$, and the [quadratic form](../../01_Linear_Algebra/07_Theoretical_Linear_Algebra/17_positive_definite_matrix.ipynb) $\mathbf{x}^\top Q\mathbf{x}$ has gradient $(Q + Q^\top)\mathbf{x}$, which simplifies to $2Q\mathbf{x}$ for symmetric $Q$. These two results are the most-used gradients in the repository: a general quadratic objective $\tfrac12\mathbf{x}^\top Q\mathbf{x} - \mathbf{b}^\top\mathbf{x}$ has gradient $Q\mathbf{x} - \mathbf{b}$ and Hessian $Q$, which is the basis of [quadratic minimization](../../../03_Optimization/01_Unconstrained_Optimization/05_quadratic_cost_functions_and_existence.ipynb), the LQR cost, and least squares.

**Properties:**
- Hessian of $\tfrac12\mathbf{x}^\top Q\mathbf{x}$ is the symmetric part of $Q$; for symmetric $Q$ it is $Q$ itself.
- Stationary point of $\tfrac12\mathbf{x}^\top Q\mathbf{x} - \mathbf{b}^\top\mathbf{x}$ solves $Q\mathbf{x} = \mathbf{b}$.

**Numerical Example:**

Take $Q = \begin{bmatrix} 4 & 1 \\ 1 & 2 \end{bmatrix}$ (symmetric). The quadratic form is

$$
\mathbf{x}^\top Q\mathbf{x} = 4x_1^2 + 2x_1x_2 + 2x_2^2,
$$

and differentiating gives

$$
\nabla(\mathbf{x}^\top Q\mathbf{x}) = \begin{bmatrix} 8x_1 + 2x_2 \\ 2x_1 + 4x_2 \end{bmatrix} = 2Q\mathbf{x} .
$$

The Hessian is the constant $2Q = \begin{bmatrix} 8 & 2 \\ 2 & 4 \end{bmatrix}$, confirming $\nabla^2(\mathbf{x}^\top Q\mathbf{x}) = 2Q$.

```python
import sympy as sp

x_1, x_2 = sp.symbols("x_1 x_2")
x = sp.Matrix([x_1, x_2])
Q = sp.Matrix([[4, 1], [1, 2]])

quadratic_form = (x.T * Q * x)[0]
gradient = sp.Matrix([sp.diff(quadratic_form, variable) for variable in x])
hessian = sp.hessian(quadratic_form, list(x))

print("xᵀQx          =", sp.expand(quadratic_form))
print("∇(xᵀQx)       =", gradient.T)
print("equals 2Qx    :", sp.simplify(gradient - 2 * Q * x) == sp.zeros(2, 1))
print("∇²(xᵀQx) = 2Q ="); sp.pprint(hessian)
```

**Equivalent CasADi implementation:**

`ca.gradient` and `ca.hessian` return $2Q\mathbf{x}$ and $2Q$ for the quadratic form.

```python
import casadi as ca

x = ca.SX.sym("x", 2)
Q = ca.DM([[4, 1], [1, 2]])

quadratic_form = ca.bilin(Q, x, x)
hessian, gradient = ca.hessian(quadratic_form, x)

print("∇(xᵀQx) =", gradient)
print("∇²(xᵀQx) = 2Q =", hessian)
```
