### 1.3.2.7. Mean Value Theorem

$$
f \text{ continuous on } [a,b],\ \text{differentiable on } (a,b)
\;\Longrightarrow\;
\exists\, c \in (a,b) : f'(c) = \frac{f(b) - f(a)}{b - a} .
$$

**Explanation:**

The Mean Value Theorem guarantees an interior point where the instantaneous rate of change equals the average rate of change over $[a,b]$ — geometrically, a tangent line parallel to the secant joining the endpoints. Its special case $f(a) = f(b)$ is Rolle's theorem. The theorem is the bridge from a bound on the derivative to a bound on the function: if $|f'| \le L$ everywhere, then $|f(b) - f(a)| \le L|b - a|$, the Lipschitz/descent-type estimate that convergence proofs for [gradient methods](../../../03_Optimization/01_Unconstrained_Optimization/10_gradient_method_convergence.ipynb) rely on.

**Numerical Example:**

Let $f(x) = x^2$ on $[1, 3]$. The average rate of change is

$$
\frac{f(3) - f(1)}{3 - 1} = \frac{9 - 1}{2} = 4 .
$$

Setting $f'(c) = 2c$ equal to $4$ gives $c = 2$, which lies in $(1, 3)$ — the guaranteed point. The tangent at $x = 2$ has slope $4$, matching the secant from $(1,1)$ to $(3,9)$.

```python
import sympy as sp

x = sp.symbols("x")
f = x**2
left_end, right_end = 1, 3

average_rate = (f.subs(x, right_end) - f.subs(x, left_end)) / (right_end - left_end)
mvt_points = sp.solve(sp.Eq(sp.diff(f, x), average_rate), x)
guaranteed_point = [point for point in mvt_points if left_end < point < right_end]

print("average rate of change on [1,3] :", average_rate)
print("solve f'(c) = average rate       :", mvt_points)
print("interior point c guaranteed by MVT:", guaranteed_point)
```
