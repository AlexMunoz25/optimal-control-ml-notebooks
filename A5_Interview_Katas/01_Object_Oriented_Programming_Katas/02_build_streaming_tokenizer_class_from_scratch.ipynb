### A5.1.2. Build Streaming Tokenizer Class from Scratch

**Problem:**

Build a class that reads a text stream character by character and emits tokens. A token is a contiguous run of non-whitespace characters.

**Explanation:**

A streaming tokenizer processes input one character at a time. It accumulates characters into a buffer while they are non-whitespace. When it hits whitespace or end-of-input, it yields the accumulated buffer as a token and resets.

How to build it:

1. Store an input stream (any object with a `read(1)` method) in `__init__`.
2. Keep a buffer (list of characters) as internal state.
3. Make the class iterable with `__iter__` and `__next__`.
4. In `__next__`, read one character at a time in a loop. If the character is non-whitespace, append it to the buffer. If it is whitespace and the buffer is non-empty, join the buffer into a string, clear the buffer, and return the token.
5. When `read(1)` returns an empty string (end of stream), yield any remaining buffer content, then raise `StopIteration`.

**Example:**

Given input `"hello world  foo"`, the tokenizer yields `"hello"`, `"world"`, `"foo"`.

```python
import io


class StreamingTokenizer:
    def __init__(self, stream):
        self.stream = stream
        self.buffer = []
        self.exhausted = False

    def __iter__(self):
        return self

    def __next__(self):
        while True:
            char = self.stream.read(1)

            if not char:
                self.exhausted = True
                if self.buffer:
                    token = "".join(self.buffer)
                    self.buffer.clear()
                    return token
                raise StopIteration

            if char.isspace():
                if self.buffer:
                    token = "".join(self.buffer)
                    self.buffer.clear()
                    return token
            else:
                self.buffer.append(char)


stream = io.StringIO("hello world  foo  bar")
tokenizer = StreamingTokenizer(stream)

for token in tokenizer:
    print(repr(token))
```
