### A5.1.4. Build Object Pool Class from Scratch

**Problem:**

Build a class that pre-allocates a fixed number of reusable objects. Clients check out objects from the pool and return them when done, avoiding repeated allocation.

**Explanation:**

An object pool keeps a collection of pre-created objects. Instead of creating new objects each time, clients borrow one from the pool and return it after use. This reduces allocation overhead for expensive-to-create objects.

How to build it:

1. Accept a factory callable and a pool size in `__init__`.
2. Pre-create all objects by calling the factory, store them in a list (the available pool).
3. `acquire()`: pop an object from the available list and return it. If the pool is empty, raise an exception or block.
4. `release(obj)`: put the object back into the available list.

**Example:**

A pool of 3 database connections. A client acquires one, uses it, and releases it back. The next client gets the same connection object without creating a new one.

```python
class ObjectPool:
    def __init__(self, factory, size):
        self.available = [factory() for _ in range(size)]

    def acquire(self):
        return self.available.pop()

    def release(self, obj):
        self.available.append(obj)


connection_id = 0


def create_connection():
    global connection_id
    connection_id += 1
    return {"id": connection_id, "status": "open"}


pool = ObjectPool(create_connection, size=3)
print(f"Pool created with {len(pool.available)} connections")

conn = pool.acquire()
print(f"Acquired connection: {conn}")
print(f"Pool size after acquire: {len(pool.available)}")

pool.release(conn)
print(f"Pool size after release: {len(pool.available)}")
```
