### 1.3.6.5. Divergence

$$
\operatorname{div}\mathbf{F} = \nabla \cdot \mathbf{F}
= \frac{\partial P}{\partial x} + \frac{\partial Q}{\partial y} + \frac{\partial R}{\partial z}
= \operatorname{tr}(J_{\mathbf{F}}) .
$$

**Explanation:**

The divergence of a [vector field](./04_vector_fields_and_gradient_fields.ipynb) is the scalar $\nabla\cdot\mathbf{F}$, the sum of the diagonal partial derivatives — equivalently the trace of the field's [Jacobian](../05_Multivariable_Differential_Calculus/07_jacobian.ipynb). It measures the net outflow per unit volume at a point: positive divergence marks a *source*, negative a *sink*, and zero an incompressible (volume-preserving) flow. Divergence is the local form of conservation laws — mass, probability, or energy transport — and is the operator the continuity and Fokker–Planck equations are built from.

**Properties:**
- $\operatorname{div}\mathbf{F} = \operatorname{tr}(J_{\mathbf{F}})$, the sum of the field's diagonal partials.
- A divergence-free field ($\nabla\cdot\mathbf{F} = 0$) conserves volume along its flow.

**Intuition:**

These fields are divergence-free: as much flow enters any small region as leaves it, so there is no net source or sink.

**Numerical Example:**

For the radial field $\mathbf{F} = (x, y)$: $P = x$, $Q = y$, so

$$
\operatorname{div}\mathbf{F} = \frac{\partial x}{\partial x} + \frac{\partial y}{\partial y} = 1 + 1 = 2 > 0 ,
$$

a uniform source — the arrows spread outward everywhere. For the rotational field $\mathbf{G} = (-y, x)$:

$$
\operatorname{div}\mathbf{G} = \frac{\partial(-y)}{\partial x} + \frac{\partial(x)}{\partial y} = 0 + 0 = 0 ,
$$

incompressible — the flow circulates without net outflow.

```python
import sympy as sp

x, y = sp.symbols("x y")

def divergence(field, variables):
    return sum(sp.diff(component, variable) for component, variable in zip(field, variables))

radial_field = (x, y)
rotational_field = (-y, x)

print("div (x, y)   =", divergence(radial_field, (x, y)), " (source)")
print("div (-y, x)  =", divergence(rotational_field, (x, y)), " (incompressible)")
```

**Equivalent CasADi implementation:**

CasADi computes the divergence as the trace of the field's Jacobian.

```python
import casadi as ca

position = ca.SX.sym("p", 2)
radial_field = ca.vertcat(position[0], position[1])
divergence_expression = ca.trace(ca.jacobian(radial_field, position))

print("div (x, y) = tr(J) =", float(ca.evalf(divergence_expression)))
```
