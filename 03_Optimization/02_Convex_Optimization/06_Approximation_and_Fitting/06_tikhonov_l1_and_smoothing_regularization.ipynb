### 3.2.6.6. Tikhonov, L1, and Smoothing Regularization


$$\|Ax-b\|_2^2+\lambda\|Dx\|_2^2$$


**Explanation:**

Regularization adds structure to an approximation problem by penalizing coefficient size, sparsity, or roughness. Tikhonov, l1, and smoothing penalties express different tradeoffs between fitting the data and controlling the solution.

The residual term measures data fit, while $Dx$ applies a [matrix](../../../01_Foundations/01_Linear_Algebra/03_Matrix/01_matrix_definition.ipynb) $D$ before measuring roughness or size. The parameter $\lambda$ controls how strongly that regularization term influences the objective.

**Properties:**
- The same residual vector can have different objective values.
- The infinity [norm](../../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb) is the maximum absolute residual.
- Norm and penalty choices encode robustness or smoothness preferences.

**Intuition:**

The example evaluates three common residual measures on one vector.


**Numerical Example:**

Let
$$
r=\begin{bmatrix}1\\-2\\3\end{bmatrix}.
$$

The squared two-norm is
$$
\|r\|_2^2=1^2+(-2)^2+3^2=1+4+9=14.
$$

The one-norm is
$$
\|r\|_1=|1|+|-2|+|3|=1+2+3=6.
$$

The infinity norm is
$$
\|r\|_\infty=\max\{1,2,3\}=3.
$$


```python
import sympy as sp

residual = sp.Matrix([1, -2, 3])
squared_l2 = sum(entry ** 2 for entry in residual)
l1_norm = sum(abs(entry) for entry in residual)
linfinity_norm = max(abs(entry) for entry in residual)

print("squared_l2 =", squared_l2)
print("l1_norm =", l1_norm)
print("linfinity_norm =", linfinity_norm)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

residual = ca.DM([1, -2, 3])
squared_l2 = ca.dot(residual, residual)
l1_norm = ca.norm_1(residual)
linfinity_norm = ca.norm_inf(residual)

print("squared_l2 =", squared_l2)
print("l1_norm =", l1_norm)
print("linfinity_norm =", linfinity_norm)
```
