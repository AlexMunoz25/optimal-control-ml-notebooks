### 1.4.6.5. Convolution Theorem

$$
(f * g)(t) = \int_0^t f(\tau)\,g(t - \tau)\,d\tau,
\qquad
\mathcal{L}\{f * g\} = F(s)\,G(s) .
$$

**Explanation:**

The **convolution** $f * g$ blends two functions by sliding one across the other and integrating the overlap. The **convolution theorem** is the operational gem: the transform of a convolution is the *product* of the transforms. Run in reverse, it inverts a product $F(s)G(s)$ as the convolution $f * g$ without a partial-fraction expansion. In systems terms it is fundamental — the output of a linear time-invariant system is the convolution of its input with the system's [impulse response](./06_dirac_delta_and_transfer_functions.ipynb), which is exactly multiplication by the transfer function in the $s$-domain.

**Properties:**
- Convolution is commutative: $f * g = g * f$.
- Transform of an integral: $\mathcal{L}\left\{\int_0^t f(\tau)\,d\tau\right\} = \dfrac{F(s)}{s}$ (the case $g = 1$).

**Numerical Example:**

Evaluate $\mathcal{L}\left\{\displaystyle\int_0^t e^{\tau}\sin(t - \tau)\,d\tau\right\}$, which is $\mathcal{L}\{e^t * \sin t\}$.

**By the theorem,** it is the product of the individual transforms:

$$
\mathcal{L}\{e^t * \sin t\} = \mathcal{L}\{e^t\}\,\mathcal{L}\{\sin t\} = \frac{1}{s-1}\cdot\frac{1}{s^2+1} = \frac{1}{(s-1)(s^2+1)} .
$$

**Direct check.** Evaluating the convolution integral gives $e^t * \sin t = \tfrac12\big(e^t - \cos t - \sin t\big)$, and transforming that combination returns the same $\dfrac{1}{(s-1)(s^2+1)}$ — confirming the theorem turns the integral into a simple product.

```python
import sympy as sp

t, s, tau = sp.symbols("t s tau", positive=True)

convolution = sp.integrate(sp.exp(tau) * sp.sin(t - tau), (tau, 0, t))
transform_of_product = (sp.laplace_transform(sp.exp(t), t, s, noconds=True)
                        * sp.laplace_transform(sp.sin(t), t, s, noconds=True))
transform_of_convolution = sp.laplace_transform(convolution, t, s, noconds=True)

print("e^t * sin t =", sp.simplify(convolution))
print("L{e^t} L{sin t} =", sp.simplify(transform_of_product))
print("L{e^t * sin t} =", sp.factor(transform_of_convolution))
```
