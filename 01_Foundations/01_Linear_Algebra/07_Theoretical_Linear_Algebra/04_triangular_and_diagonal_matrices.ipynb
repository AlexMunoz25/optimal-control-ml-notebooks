### 1.1.7.4. Triangular and Diagonal Matrices


$$
U_{ij} = 0 \quad \text{if } i > j
\qquad
L_{ij} = 0 \quad \text{if } i < j
\qquad
D_{ij} = 0 \quad \text{if } i \neq j
$$


**Explanation:**

A triangular matrix has all entries on one side of the main diagonal equal to zero. An upper triangular matrix $U$ satisfies $U_{ij} = 0$ for all $i > j$; a lower triangular matrix $L$ satisfies $L_{ij} = 0$ for all $i < j$. A diagonal matrix satisfies both conditions simultaneously. These structures share a key spectral property—their eigenvalues are exactly the diagonal entries—which makes determinants, traces, and inverses easy to compute. They arise as factors in [LU decomposition](./18_lu_decomposition.ipynb), [Cholesky decomposition](./19_cholesky_decomposition.ipynb), and [QR decomposition](./20_qr_decomposition.ipynb), where their structure enables forward and backward substitution.

**Properties:**

- Eigenvalues equal the diagonal entries.
- Determinant equals the product of the diagonal entries.
- The product of two upper (or two lower) triangular matrices is triangular of the same type.
- The inverse of a nonsingular triangular matrix is triangular of the same type.
- A diagonal matrix $\text{diag}(d_1, \ldots, d_n)$ is invertible if and only if all $d_i \neq 0$, with inverse $\text{diag}(1/d_1, \ldots, 1/d_n)$.

**Example:**

For the upper triangular matrix

$$
U =
\begin{bmatrix}
2 & 3 & 1 \\
0 & 4 & 5 \\
0 & 0 & 6
\end{bmatrix},
$$

the eigenvalues are $\lambda = 2, 4, 6$ and $\det(U) = 2 \cdot 4 \cdot 6 = 48$.


```python
import sympy as sp

upper_triangular = sp.Matrix([[2, 3, 1], [0, 4, 5], [0, 0, 6]])
diagonal = sp.Matrix([[2, 0, 0], [0, 4, 0], [0, 0, 6]])

print("eigenvalues =", list(upper_triangular.eigenvals().keys()))
print("determinant =", upper_triangular.det())
print("diagonal inverse =")
sp.pprint(diagonal.inv())
```
