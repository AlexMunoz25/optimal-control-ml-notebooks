### 1.3.4.6. Quadratic Approximation and Curvature

$$
f(x) \approx f(a) + f'(a)(x - a) + \tfrac{1}{2} f''(a)(x - a)^{2} = q(x) .
$$

**Explanation:**

The quadratic (second-order Taylor) approximation adds the curvature term $\tfrac12 f''(a)(x-a)^2$ to the [linear approximation](../02_Single_Variable_Differentiation/10_linear_approximation.ipynb), producing the best parabola matching $f$ in value, slope, and [second derivative](../02_Single_Variable_Differentiation/09_convexity_and_second_derivative_test.ipynb) at $a$. The sign of $f''(a)$ decides whether that parabola opens up (local min) or down (local max), which is the second-derivative test, and its magnitude sets how sharply the function bends. This one-variable model is the direct precursor of the multivariable [second-order Taylor expansion](../05_Multivariable_Differential_Calculus/10_second_order_taylor_expansion.ipynb) and the quadratic models that Newton-type methods minimize at each step.

**Properties:**
- $q$ matches $f$ to second order: $q(a)=f(a)$, $q'(a)=f'(a)$, $q''(a)=f''(a)$, with error $O\big((x-a)^3\big)$.
- A Newton step minimizes this quadratic model: $x_{\text{new}} = a - f'(a)/f''(a)$ when $f''(a) > 0$.

**Numerical Example:**

Expand $f(x) = \cos x$ about $a = 0$. With $f(0) = 1$, $f'(0) = 0$, $f''(0) = -1$,

$$
q(x) = 1 + 0\cdot x + \tfrac12(-1)x^2 = 1 - \frac{x^2}{2} .
$$

At $x = 0.3$: $q(0.3) = 1 - 0.045 = 0.955$, versus $\cos(0.3) = 0.95534\ldots$ — an error of $\approx 3\times 10^{-4}$. The negative curvature $f''(0) = -1$ shows the cosine bends downward at its peak, consistent with $x = 0$ being a local maximum.

```python
import sympy as sp

x = sp.symbols("x")
f = sp.cos(x)
center, evaluation_point = 0, sp.Rational(3, 10)

value = f.subs(x, center)
slope = sp.diff(f, x).subs(x, center)
curvature = sp.diff(f, x, 2).subs(x, center)
quadratic_model = value + slope * (x - center) + sp.Rational(1, 2) * curvature * (x - center)**2

approximate_value = quadratic_model.subs(x, evaluation_point)
true_value = f.subs(x, evaluation_point)

print("q(x) =", quadratic_model)
print("f(0), f'(0), f''(0) =", value, slope, curvature)
print("q(0.3)      =", float(approximate_value))
print("cos(0.3)    =", sp.N(true_value, 8))
print("error       =", sp.N(true_value - approximate_value))
```

**Equivalent CasADi implementation:**

CasADi builds the same second-order model from the gradient and Hessian (here $1\times 1$) of $f$ at the center.

```python
import casadi as ca

x = ca.SX.sym("x")
f = ca.cos(x)
hessian, gradient = ca.hessian(f, x)
model_data = ca.Function("model", [x], [f, gradient, hessian])

center = 0.0
value, slope, curvature = [float(item) for item in model_data(center)]
evaluation_point = 0.3
quadratic_estimate = value + slope * (evaluation_point - center) + 0.5 * curvature * (evaluation_point - center)**2

print("f(0), f'(0), f''(0) =", value, slope, curvature)
print("q(0.3)   =", quadratic_estimate)
print("cos(0.3) =", float(ca.cos(ca.DM(0.3))))
```
