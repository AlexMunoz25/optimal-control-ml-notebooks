### 3.4.1. Linear Programming (LP)

$$\min_x c^Tx\quad\text{subject to}\quad Ax\preceq b$$


**Explanation:**

Linear programs minimize a linear objective over a polyhedron. Feasibility is described by affine inequalities, and optimality can be understood geometrically by sliding objective hyperplanes until they first touch the feasible set.

The vector $x$ is the decision variable and $c^Tx$ is a linear objective formed by an [inner product](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/02_inner_product_space.ipynb). The matrix inequality $Ax\preceq b$ represents a finite set of affine inequality constraints.

**Properties:**
- The feasible set is an intersection of halfspaces.
- The objective has affine level sets.
- An optimum, when finite and attained, occurs on a face of the polyhedron.

**Intuition:**

The figure above shows linear objective geometry over a feasible polygon; the calculation compares feasible vertices.

<img src="../../Figures/030305_bv_fig_4_4_linear_program_geometry.jpeg" alt="Boyd and Vandenberghe Figure 4.4: geometric interpretation of a linear program" style="width: 60%; height: auto;">


**Numerical Example:**

The figure above shows the same vertex-comparison geometry for a linear program.

Let
$$
c=\begin{bmatrix}1\\2\end{bmatrix}
$$
and compare feasible vertices
$$
x^{(1)}=\begin{bmatrix}1\\0\end{bmatrix},\quad
x^{(2)}=\begin{bmatrix}0\\1\end{bmatrix},\quad
x^{(3)}=\begin{bmatrix}1\\1\end{bmatrix}.
$$

The objective values are
$$
c^Tx^{(1)}=1,
\qquad
c^Tx^{(2)}=2,
\qquad
c^Tx^{(3)}=3.
$$

Among these feasible candidates, the smallest value is $1$ at $x^{(1)}$.


```python
import sympy as sp

cost_vector = sp.Matrix([1, 2])
candidate_points = [sp.Matrix([1, 0]), sp.Matrix([0, 1]), sp.Matrix([1, 1])]
costs = [(cost_vector.T * candidate)[0] for candidate in candidate_points]
optimizer = candidate_points[costs.index(min(costs))]

print("optimizer =", optimizer)
print("costs =", costs)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

cost_vector = ca.DM([1, 2])
candidate_points = [ca.DM([1, 0]), ca.DM([0, 1]), ca.DM([1, 1])]
costs = [float(ca.dot(cost_vector, candidate)) for candidate in candidate_points]
optimizer = candidate_points[costs.index(min(costs))]

print("optimizer =", optimizer)
print("costs =", costs)
```

**Feasible-set geometry:**

A linear program is minimized over a polyhedron, the intersection of finitely many halfspaces. In two dimensions this feasible set is a filled polygon with straight edges, and in three dimensions a bounded polyhedral face such as the simplex $\{x\succeq0,\ \mathbf{1}^Tx=1\}$ with vertices $(1,0,0)$, $(0,1,0)$, and $(0,0,1)$. The cells below rebuild both feasible sets directly from their defining vertices.

<img src="../../Figures/030305_lp_feasible_polyhedron_2d.png" alt="Linear program feasible polyhedron in two dimensions" style="width: 40%; height: auto;">
<img src="../../Figures/030305_lp_feasible_polyhedron_3d.png" alt="Linear program feasible polyhedron in three dimensions" style="width: 46%; height: auto;">

```python
import numpy as np
import matplotlib.pyplot as plt

feasible_vertices = np.array([[0, 0], [0, 1], [0.25, 1], [1, 0]])

figure, axis = plt.subplots(figsize=(5, 5))
axis.fill(feasible_vertices[:, 0], feasible_vertices[:, 1], color="#E08E2F", edgecolor="#9c5a1c", linewidth=1.5)
axis.set_xlim(-0.08, 1.12)
axis.set_ylim(-0.08, 1.12)
axis.set_aspect("equal")
axis.set_xticks([])
axis.set_yticks([])
for spine in axis.spines.values():
    spine.set_color("#cfcfcf")
figure.tight_layout()
figure.savefig("../../Figures/030305_lp_feasible_polyhedron_2d.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

simplex_vertices = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]])

figure = plt.figure(figsize=(5, 5))
axis = figure.add_subplot(111, projection="3d")
axis.add_collection3d(Poly3DCollection([simplex_vertices], facecolor="#E08E2F", edgecolor="#9c5a1c", linewidth=1.2))
axis.set_xlim(0, 1)
axis.set_ylim(0, 1)
axis.set_zlim(0, 1)
axis.view_init(elev=20, azim=-120)
for pane in (axis.xaxis.pane, axis.yaxis.pane, axis.zaxis.pane):
    pane.set_facecolor((1, 1, 1, 0))
    pane.set_edgecolor("#cfcfcf")
axis.set_xticklabels([])
axis.set_yticklabels([])
axis.set_zticklabels([])
axis.tick_params(length=0)
figure.savefig("../../Figures/030305_lp_feasible_polyhedron_3d.png", dpi=150, bbox_inches="tight")
plt.show()
```
