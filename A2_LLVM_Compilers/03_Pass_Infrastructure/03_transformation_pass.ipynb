### A2.3.3. Transformation Pass

> *A transformation pass modifies the IR to improve performance, reduce code size, or canonicalize its structure. It consumes analysis results and reports which analyses it preserves.*

**Explanation:**

A **transformation pass** mutates the IR. After running, it returns a `PreservedAnalyses` set that tells the analysis manager which cached results are still valid.

**Common Return Values:**

- `PreservedAnalyses::all()` — no IR was changed; all analyses remain valid.
- `PreservedAnalyses::none()` — all analyses are invalidated.
- Selective preservation — `PA.preserve<DominatorTreeAnalysis>()` keeps specific analyses.

**Built-in Transformation Passes:**

- `InstCombine` — algebraic simplification and canonicalization of instructions.
- `SimplifyCFG` — merge blocks, eliminate dead branches, thread jumps.
- `GVN` — global value numbering; eliminates redundant computations.
- `LICM` — loop-invariant code motion; hoists invariant computations out of loops.
- `SROA` — scalar replacement of aggregates; promotes struct/array allocas to SSA values.

**Example:**

```cpp
struct MyTransform : public PassInfoMixin<MyTransform> {
  PreservedAnalyses run(Function &F, FunctionAnalysisManager &AM) {
    auto &DT = AM.getResult<DominatorTreeAnalysis>(F);
    bool changed = false;
    // ... mutate IR using dominator tree ...
    if (!changed)
      return PreservedAnalyses::all();
    PreservedAnalyses PA;
    PA.preserve<DominatorTreeAnalysis>();
    return PA;
  }
};
```

````python
transform_passes = [
    ("instcombine", "Algebraic simplification", "a + 0 -> a, icmp folding"),
    ("simplifycfg", "CFG simplification", "dead branch removal, block merging"),
    ("gvn", "Global value numbering", "redundant load/computation elimination"),
    ("licm", "Loop-invariant code motion", "hoist invariant ops out of loops"),
    ("sroa", "Scalar replacement of aggregates", "promote allocas to SSA"),
    ("dse", "Dead store elimination", "remove stores never read"),
    ("adce", "Aggressive dead code elimination", "remove dead instructions"),
]

print("Key LLVM Transformation Passes:")
print(f"{'Pass':<16} {'Purpose':<32} {'Example'}")
print("-" * 80)
for name, purpose, example in transform_passes:
    print(f"{name:<16} {purpose:<32} {example}")
````
