### 1.4.4.1. Preliminary Theory of Linear Equations

$$
y = \underbrace{c_1 y_1 + \cdots + c_n y_n}_{y_c\ (\text{complementary})} + \underbrace{y_p}_{\text{particular}},
\qquad
W(y_1, \ldots, y_n) =
\begin{vmatrix}
y_1 & \cdots & y_n \\
\vdots & & \vdots \\
y_1^{(n-1)} & \cdots & y_n^{(n-1)}
\end{vmatrix} \neq 0 .
$$

**Explanation:**

For a linear $n$th-order equation $L(y) = g(x)$, solutions have rich structure. The **superposition principle** says any linear combination of solutions of the homogeneous equation $L(y) = 0$ is again a solution. A set of $n$ solutions is a **fundamental set** when it is linearly independent, which the **Wronskian** [determinant](../../01_Linear_Algebra/03_Matrix/10_matrix_determinant.ipynb) detects: $W \neq 0$ on the interval. The general solution of $L(y) = 0$ is then the **complementary function** $y_c = c_1 y_1 + \cdots + c_n y_n$, and the general solution of the nonhomogeneous equation is $y = y_c + y_p$, where $y_p$ is any particular solution. So every later method splits into "find $y_c$" and "find $y_p$."

**Properties:**
- A homogeneous linear equation always has the trivial solution $y = 0$; nontrivial solutions span an $n$-dimensional space.
- $y = y_c + y_p$: solve the associated homogeneous equation, then add one particular solution.

**Numerical Example:**

Consider $y''' - 6y'' + 11y' - 6y = 3x$.

**Complementary function.** The functions $y_1 = e^x$, $y_2 = e^{2x}$, $y_3 = e^{3x}$ each solve the homogeneous equation. Their Wronskian is

$$
W(e^x, e^{2x}, e^{3x}) =
\begin{vmatrix}
e^x & e^{2x} & e^{3x} \\
e^x & 2e^{2x} & 3e^{3x} \\
e^x & 4e^{2x} & 9e^{3x}
\end{vmatrix} = 2e^{6x} \neq 0,
$$

so they form a fundamental set and $y_c = c_1 e^x + c_2 e^{2x} + c_3 e^{3x}$.

**Particular solution.** Substituting a linear trial $y_p = ax + b$ gives $y_p = -\tfrac{1}{2}x - \tfrac{11}{12}$. Therefore the general solution is

$$
y = y_c + y_p = c_1 e^x + c_2 e^{2x} + c_3 e^{3x} - \tfrac{1}{2}x - \tfrac{11}{12}.
$$

```python
import sympy as sp

x = sp.symbols("x")
y = sp.Function("y")

basis = [sp.exp(x), sp.exp(2 * x), sp.exp(3 * x)]
wronskian = sp.Matrix([[sp.diff(function, x, order) for function in basis] for order in range(3)]).det()

homogeneous = sp.dsolve(sp.Eq(y(x).diff(x, 3) - 6 * y(x).diff(x, 2) + 11 * y(x).diff(x) - 6 * y(x), 0), y(x))
nonhomogeneous = sp.dsolve(sp.Eq(y(x).diff(x, 3) - 6 * y(x).diff(x, 2) + 11 * y(x).diff(x) - 6 * y(x), 3 * x), y(x))

print("Wronskian W(e^x, e^2x, e^3x) =", sp.simplify(wronskian), " (nonzero -> fundamental set)")
print("complementary y_c :", homogeneous.rhs)
print("general y = y_c+y_p:", nonhomogeneous.rhs)
```
