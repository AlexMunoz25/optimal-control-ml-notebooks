### 1.3.5.9. Hessian Matrix

$$
H(\mathbf{x}) = \nabla^2 f(\mathbf{x}) =
\begin{bmatrix}
\dfrac{\partial^2 f}{\partial x_1^2} & \cdots & \dfrac{\partial^2 f}{\partial x_1 \partial x_n} \\
\vdots & \ddots & \vdots \\
\dfrac{\partial^2 f}{\partial x_n \partial x_1} & \cdots & \dfrac{\partial^2 f}{\partial x_n^2}
\end{bmatrix}.
$$

**Explanation:**

The Hessian is the matrix of all second-order partial derivatives — the multivariable analogue of the [second derivative](../02_Single_Variable_Differentiation/06_higher_order_derivatives.ipynb) and the [Jacobian](./07_jacobian.ipynb) of the gradient. When the second partials are continuous, mixed partials are equal ($f_{xy} = f_{yx}$), so the Hessian is a [symmetric matrix](../../01_Linear_Algebra/07_Theoretical_Linear_Algebra/16_symmetric_matrix.ipynb). It encodes the local curvature of $f$, and its [definiteness](../../01_Linear_Algebra/07_Theoretical_Linear_Algebra/18_matrix_definiteness.ipynb) classifies critical points and sets the conditioning of [Newton's method](../../../03_Optimization/05_Unconstrained_Algorithms/11_newtons_method.ipynb).

**Properties:**
- Symmetry (Clairaut/Schwarz): $f_{x_i x_j} = f_{x_j x_i}$ for $f$ with continuous second partials.
- Positive definite $H$ at a critical point $\Rightarrow$ local minimum; negative definite $\Rightarrow$ local maximum.

**Numerical Example:**

For $f(x, y) = x^2 + xy + y^2$, the first partials are $f_x = 2x + y$, $f_y = x + 2y$. Differentiating again,

$$
f_{xx} = 2,\quad f_{yy} = 2,\quad f_{xy} = f_{yx} = 1
\;\Rightarrow\;
H = \begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix}.
$$

The Hessian is constant (because $f$ is quadratic) and symmetric. Its eigenvalues are $3$ and $1$, both positive, so $H$ is positive definite and $f$ is convex with a unique minimum.

```python
import sympy as sp

x, y = sp.symbols("x y")
f = x**2 + x * y + y**2
variables = [x, y]

hessian = sp.hessian(f, variables)
is_symmetric = hessian == hessian.T
eigenvalues = list(hessian.eigenvals().keys())

print("Hessian ="); sp.pprint(hessian)
print("symmetric:", is_symmetric)
print("eigenvalues:", eigenvalues, "-> positive definite (convex)")
```

**Equivalent CasADi implementation:**

`ca.hessian` returns the same second-derivative matrix (and the gradient as a by-product).

```python
import casadi as ca

variables = ca.SX.sym("x", 2)
f = variables[0]**2 + variables[0] * variables[1] + variables[1]**2
hessian, gradient = ca.hessian(f, variables)

print("Hessian =", hessian)
```
