### 1.4.1.4. Existence and Uniqueness of Solutions

$$
\text{If } f(x, y) \text{ and } \frac{\partial f}{\partial y} \text{ are continuous on a rectangle } R \ni (x_0, y_0),
$$

$$
\text{then } \frac{dy}{dx} = f(x, y),\ y(x_0) = y_0 \text{ has a unique solution on some interval } (x_0 - h,\, x_0 + h).
$$

**Explanation:**

Before solving an IVP it is worth knowing whether a solution exists and whether it is the only one. The theorem gives *sufficient* conditions: if $f(x,y)$ is [continuous](../../03_Calculus/01_Limits_and_Continuity/04_continuity.ipynb) on a rectangle around $(x_0, y_0)$ a solution **exists**, and if in addition the [partial derivative](../../03_Calculus/05_Multivariable_Differential_Calculus/02_partial_derivatives.ipynb) $\partial f / \partial y$ is continuous there, the solution is **unique** near $x_0$. These conditions are sufficient but not necessary, and uniqueness is only local — guaranteed on some (possibly small) interval about $x_0$. When $\partial f/\partial y$ fails to be continuous, uniqueness can break down.

**Intuition:**

When the hypotheses fail, the same initial point can lie on more than one solution curve.

       alt="Two solution curves, y = x^4/16 and y = 0, both passing through the origin"
       width="220">
</p>

**Numerical Example:**


Consider the IVP $\dfrac{dy}{dx} = x\,y^{1/2}$, $y(0) = 0$, with $f(x,y) = x\,y^{1/2}$. It has **two** solutions.

*First solution* $y = \tfrac{1}{16}x^4$: the derivative is $y' = \tfrac{1}{4}x^3$, and since the nonnegative root is $y^{1/2} = \big(\tfrac{1}{16}x^4\big)^{1/2} = \tfrac14 x^2$,
$$
x\,y^{1/2} = x\cdot\tfrac14 x^2 = \tfrac14 x^3 = y', \qquad y(0) = 0. \;\checkmark
$$

*Second solution* $y \equiv 0$: then $y' = 0$ and $x\,y^{1/2} = x\cdot 0 = 0$, with $y(0) = 0$. $\;\checkmark$

Both satisfy the same IVP, so the solution is **not unique**. The reason is visible in
$$
\frac{\partial f}{\partial y} = \frac{\partial}{\partial y}\big(x\,y^{1/2}\big) = \frac{x}{2\,y^{1/2}},
$$
which is discontinuous along $y = 0$, so the theorem does not apply there. By contrast, at a point like $(2, 1)$ both $f$ and $\partial f/\partial y$ are continuous (since $y > 0$), so through $(2,1)$ the IVP has a unique solution.

```python
import sympy as sp

x, y = sp.symbols("x y")
f = x * sp.sqrt(y)

first_solution = x**4 / 16
left_hand_side = sp.diff(first_solution, x)
nonnegative_root = x**2 / 4
right_hand_side = x * nonnegative_root

partial_derivative = sp.diff(f, y)

print("solution 1: y = x^4/16")
print("  y'(x)              =", left_hand_side)
print("  x*y^(1/2)          =", right_hand_side, "(nonnegative root y^(1/2) = x^2/4)")
print("  satisfies the ODE  :", sp.simplify(left_hand_side - right_hand_side) == 0)
print("solution 2: y = 0 -> y' = 0 and x*0^(1/2) = 0, satisfies the ODE")
print("partial df/dy        =", partial_derivative, "-> discontinuous at y = 0")
```
