### 3.2.6.3. Penalty Function Approximation


$$\min_x\sum_i\phi(a_i^Tx-b_i)$$


**Explanation:**

Penalty approximation replaces a plain residual [norm](../../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb) with a chosen convex penalty. The penalty shape determines how small and large residuals are weighted while preserving convexity for affine residuals.

Each residual $a_i^Tx-b_i$ is an affine scalar expression in the decision vector $x$. The penalty function $\phi$ assigns a cost to that residual, and the objective sums those residual costs.

**Properties:**
- The same residual vector can have different objective values.
- The infinity norm is the maximum absolute residual.
- Norm and penalty choices encode robustness or smoothness preferences.

**Intuition:**

The example evaluates three common residual measures on one vector.

<img src="../../../Figures/03020603_bv_fig_6_1_penalty_functions.jpeg" alt="Boyd and Vandenberghe Figure 6.1: common penalty functions" style="width: 60%; height: auto;">


**Numerical Example:**

The figure above gives the visual version of the same calculation.

Let
$$
r=\begin{bmatrix}1\\-2\\3\end{bmatrix}.
$$

The squared two-norm is
$$
\|r\|_2^2=1^2+(-2)^2+3^2=1+4+9=14.
$$

The one-norm is
$$
\|r\|_1=|1|+|-2|+|3|=1+2+3=6.
$$

The infinity norm is
$$
\|r\|_\infty=\max\{1,2,3\}=3.
$$


```python
import sympy as sp

residual = sp.Matrix([1, -2, 3])
squared_l2 = sum(entry ** 2 for entry in residual)
l1_norm = sum(abs(entry) for entry in residual)
linfinity_norm = max(abs(entry) for entry in residual)

print("squared_l2 =", squared_l2)
print("l1_norm =", l1_norm)
print("linfinity_norm =", linfinity_norm)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

residual = ca.DM([1, -2, 3])
squared_l2 = ca.dot(residual, residual)
l1_norm = ca.norm_1(residual)
linfinity_norm = ca.norm_inf(residual)

print("squared_l2 =", squared_l2)
print("l1_norm =", l1_norm)
print("linfinity_norm =", linfinity_norm)
```
