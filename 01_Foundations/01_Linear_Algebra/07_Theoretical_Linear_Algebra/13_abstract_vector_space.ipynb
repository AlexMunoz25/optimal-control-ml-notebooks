### 1.1.7.13. Abstract Vector Space


$$
V
$$


**Explanation:**

An abstract vector space extends the language of vectors from $\mathbb{R}^n$ to objects such as matrices, polynomials, and functions. The central idea is that addition and scalar multiplication obey the same axioms, even when the elements no longer look like arrows. Concrete geometric versions were introduced earlier in [vector spaces](../05_Geometrical_Aspects_of_Linear_Algebra/13_vector_space.ipynb).

**Example:**

The set of polynomials of degree at most two is an abstract vector space because sums and scalar multiples of such polynomials stay in the same set.


```python
import sympy as sp

symbol_x = sp.symbols("symbol_x")
polynomial_p = 1 + 2 * symbol_x
polynomial_q = 3 - symbol_x + symbol_x**2

print("polynomial_sum =", sp.expand(polynomial_p + polynomial_q))
print("scalar_multiple =", sp.expand(2 * polynomial_q))
```
