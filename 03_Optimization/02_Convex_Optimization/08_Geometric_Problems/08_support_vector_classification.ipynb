### 3.2.8.8. Support Vector Classification


$$y_i(a^Tx_i-b)\geq1-u_i,
\qquad u_i\geq0$$


**Explanation:**

Support vector classification maximizes margin while allowing slack variables when exact separation is not possible. The soft-margin constraints are convex because they are affine inequalities with nonnegative slack. Feasibility requires the slack to cover any margin shortfall.

The label $y_i$ is usually $+1$ or $-1$, so $y_i(a^Tx_i-b)$ is the signed margin. The slack variable $u_i$ allows a margin violation, and the constraint $u_i\geq0$ prevents negative slack.

**Properties:**
- Zero slack means the margin constraint is satisfied.
- Positive slack measures violation.
- The hinge-loss form appears after eliminating slack variables.

**Intuition:**

The example computes the smallest slack for one point with insufficient margin.


**Numerical Example:**

For a positive sample, suppose the score is
$$
a^Tx-b=\frac12.
$$

The soft-margin constraint is
$$
a^Tx-b\geq1-u,\qquad u\geq0.
$$

Substitute the score:
$$
\frac12\geq1-u.
$$

Rearrange:
$$
u\geq1-\frac12=\frac12.
$$

The smallest feasible slack is $u=1/2$.


```python
import sympy as sp

score = sp.Rational(1, 2)
slack = 1 - score

print("slack =", slack)
```
