### 3.6.12. Barrier Method


$$t\leftarrow\mu t$$


**Explanation:**

The barrier method repeatedly solves centering problems while increasing the barrier parameter. Each update tightens the approximation to the original constrained problem while maintaining strict interior feasibility.

The update multiplies the barrier parameter $t$ by a factor $\mu$. Since $\mu>1$ in the barrier method, repeated updates increase the weight on the original objective relative to the barrier term.

**Properties:**
- The parameter is multiplied by a factor greater than one.
- Larger parameter values reduce the barrier gap bound.
- Each update is followed by a centering step.

**Intuition:**

The example performs two barrier-parameter updates.


**Numerical Example:**

Start with
$$
t_0=1
$$
and update factor
$$
\mu=10.
$$

After one update,
$$
t_1=\mu t_0=10\cdot1=10.
$$

After two updates,
$$
t_2=\mu t_1=10\cdot10=100.
$$

The increasing parameter makes the barrier approximation more accurate.


```python
import sympy as sp

initial_parameter = sp.Integer(1)
update_factor = sp.Integer(10)
parameter_after_one_update = update_factor * initial_parameter
parameter_after_two_updates = update_factor * parameter_after_one_update

print("parameter_after_one_update =", parameter_after_one_update)
print("parameter_after_two_updates =", parameter_after_two_updates)
```
