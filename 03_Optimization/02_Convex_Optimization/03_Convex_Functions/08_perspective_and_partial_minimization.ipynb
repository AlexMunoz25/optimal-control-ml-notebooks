### 3.2.3.8. Perspective and Partial Minimization


$$g(x,t)=tf(x/t),\qquad t>0$$


**Explanation:**

The perspective of a function rescales both its argument and value in a way that preserves convexity. Partial minimization also preserves convexity when a jointly [convex function](01_convex_function_definition_and_examples.ipynb) is minimized over some of its variables.

The perspective $g(x,t)$ first normalizes the vector $x$ by the positive scalar $t$, then multiplies the function value by $t$. The condition $t>0$ is required because the expression uses division by $t$.

**Properties:**
- The scaling coordinate must be positive.
- The map sends rays to normalized coordinates.
- Linear-fractional maps combine affine maps with this normalization.

**Intuition:**

The example divides each coordinate by the same positive scale.


**Numerical Example:**

Let
$$
z=\begin{bmatrix}6\\3\end{bmatrix},\quad t=3>0.
$$

The perspective image is
$$
P(z,t)=z/t.
$$

Compute each coordinate:
$$
\frac{z}{t}=\frac13\begin{bmatrix}6\\3\end{bmatrix}=\begin{bmatrix}2\\1\end{bmatrix}.
$$

The positive scale keeps the transformation in its domain.


```python
import sympy as sp

point = sp.Matrix([6, 3])
scale = sp.Integer(3)
image = point / scale

image
```
