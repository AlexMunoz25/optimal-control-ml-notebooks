### 1.4.8.1. Linear Systems in Matrix Form

$$
\mathbf{X}' = A\,\mathbf{X} + \mathbf{F}(t),
\qquad
\mathbf{X} = \begin{pmatrix} x_1 \\ \vdots \\ x_n \end{pmatrix},\ \
A = \begin{pmatrix} a_{11} & \cdots & a_{1n} \\ \vdots & & \vdots \\ a_{n1} & \cdots & a_{nn} \end{pmatrix}.
$$

**Explanation:**

A system of first-order linear ODEs is written compactly as $\mathbf{X}' = A\mathbf{X} + \mathbf{F}(t)$ using [matrix–vector multiplication](../../01_Linear_Algebra/03_Matrix/04_matrix_vector_multiplication.ipynb). This is precisely the **state-space form** $\dot{\mathbf{x}} = A\mathbf{x} + \cdots$ that the dynamics and control tracks build on. Any single $n$th-order linear equation converts to such a system by naming the function and its derivatives as new state variables, so the whole higher-order theory becomes a special case of systems. The homogeneous solution is a linear combination $\mathbf{X} = c_1\mathbf{X}_1 + \cdots + c_n\mathbf{X}_n$ of $n$ independent solution vectors (superposition again).

**Properties:**
- Reduction: $y^{(n)} = f(t, y, \ldots, y^{(n-1)})$ becomes a first-order system in $x_1 = y, x_2 = y', \ldots, x_n = y^{(n-1)}$.
- The eigenvalues of $A$ are the roots of the characteristic polynomial of the original equation.

**Numerical Example:**

Convert $y'' - 3y' + 2y = 0$ to a first-order system. Introduce state variables

$$
x_1 = y, \qquad x_2 = y' .
$$

Then $x_1' = y' = x_2$, and from the equation $y'' = 3y' - 2y$,

$$
x_2' = y'' = 3x_2 - 2x_1 .
$$

In matrix form,

$$
\mathbf{X}' = \begin{pmatrix} x_1' \\ x_2' \end{pmatrix} = \begin{pmatrix} 0 & 1 \\ -2 & 3 \end{pmatrix}\begin{pmatrix} x_1 \\ x_2 \end{pmatrix} = A\mathbf{X} .
$$

The eigenvalues of $A$ satisfy $\det(A - \lambda I) = \lambda^2 - 3\lambda + 2 = (\lambda-1)(\lambda-2) = 0$, i.e. $\lambda = 1, 2$ — exactly the roots of the original equation's auxiliary equation, confirming the equivalence.

```python
import sympy as sp

t = sp.symbols("t")
A = sp.Matrix([[0, 1], [-2, 3]])
characteristic = A.charpoly(sp.Symbol("lam")).as_expr()

print("companion matrix A ="); sp.pprint(A)
print("char. polynomial det(A - λI) =", sp.factor(characteristic))
print("eigenvalues (= auxiliary roots of y'' - 3y' + 2y = 0):", list(A.eigenvals().keys()))
```
