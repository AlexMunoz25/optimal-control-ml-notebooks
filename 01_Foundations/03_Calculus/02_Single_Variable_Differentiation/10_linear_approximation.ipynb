### 1.3.2.10. Linear Approximation

$$
L(x) = f(a) + f'(a)\,(x - a), \qquad f(x) \approx L(x)\ \text{ for } x \text{ near } a .
$$

**Explanation:**

The linear approximation (linearization, or tangent-line approximation) replaces $f$ near $a$ by its tangent line — the best first-order model of the function at that point. It is accurate for $x$ close to $a$ and degrades as $x$ moves away, with the error vanishing faster than $|x - a|$. This is the one-variable case of the [tangent-plane linearization](../05_Multivariable_Differential_Calculus/06_linearization_and_tangent_plane.ipynb) used to linearize dynamics around an operating point, and the first-order truncation of the [Taylor series](../04_Sequences_Series_and_Taylor_Approximation/04_taylor_and_maclaurin_series.ipynb).

**Properties:**
- $L$ matches $f$ in both value and slope at $a$: $L(a) = f(a)$ and $L'(a) = f'(a)$.
- The differential $df = f'(a)\,dx$ estimates the change $\Delta f$ for a small input change $dx$.

**Intuition:**

Near the base point the tangent line hugs the curve, so its height approximates the function value; the approximation worsens as $x$ moves away.

**Numerical Example:**

Approximate $\sqrt{4.1}$ using $f(x) = \sqrt{x}$ at $a = 4$. Here $f(4) = 2$ and $f'(x) = \tfrac{1}{2\sqrt{x}}$, so $f'(4) = \tfrac14$:

$$
L(x) = 2 + \tfrac14(x - 4) \;\Rightarrow\; L(4.1) = 2 + \tfrac14(0.1) = 2.025 .
$$

The true value is $\sqrt{4.1} = 2.024845\ldots$, so the linear model is correct to four decimals — the error $\approx 1.5\times 10^{-4}$ is tiny because $4.1$ is close to $4$.

```python
import sympy as sp

x = sp.symbols("x")
f = sp.sqrt(x)
base_point = 4
evaluation_point = sp.Rational(41, 10)

linearization = f.subs(x, base_point) + sp.diff(f, x).subs(x, base_point) * (x - base_point)
approximate_value = linearization.subs(x, evaluation_point)
true_value = f.subs(x, evaluation_point)
approximation_error = sp.N(true_value - approximate_value)

print("L(x) =", linearization)
print("L(4.1)      =", approximate_value, "=", sp.N(approximate_value))
print("sqrt(4.1)   =", sp.N(true_value, 10))
print("error       =", approximation_error)
```

**Equivalent CasADi implementation:**

CasADi builds the same first-order model from the gradient evaluated at the base point.

```python
import casadi as ca

x = ca.SX.sym("x")
f = ca.sqrt(x)
base_point = 4.0
value_and_slope = ca.Function("fs", [x], [f, ca.gradient(f, x)])

base_value, base_slope = value_and_slope(base_point)
linear_estimate = float(base_value) + float(base_slope) * (4.1 - base_point)
print("L(4.1)    =", linear_estimate)
print("sqrt(4.1) =", float(ca.sqrt(ca.DM(4.1))))
```
