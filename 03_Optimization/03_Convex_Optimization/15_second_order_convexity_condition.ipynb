### 3.3.15. Second-Order Convexity Condition

$$\nabla^2f(x)\succeq0$$


**Explanation:**

For twice differentiable functions, convexity can be checked through curvature. The foundation notebooks define the [Hessian matrix](../../01_Foundations/03_Calculus/05_Multivariable_Differential_Calculus/09_hessian_matrix.ipynb) and [matrix definiteness](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/18_matrix_definiteness.ipynb); the optimization-specific use is to certify that an objective has no hidden local minima. A positive semidefinite Hessian means every second directional derivative is nonnegative, so the function bends upward along every line in the domain.

The display says the Hessian matrix $\nabla^2 f(x)$ is positive semidefinite at each point in the domain. This is stronger than checking one point unless the Hessian is constant, as it is for a quadratic objective.

**Properties:**
- All Hessian eigenvalues must be nonnegative at every point in the domain.
- If $\nabla^2 f(x)\succ0$ everywhere, then $f$ is strictly convex.
- Convexity certificates turn local first-order optimality into global optimality.

**Intuition:**

The example checks convexity of an objective by computing its Hessian, testing positive definiteness, and then solving the first-order condition. The positive definite Hessian makes the stationary point the unique global minimizer.


**Numerical Example:**

Let
$$
f(x,y)=x^2+xy+2y^2+3x-y.
$$

The gradient is
$$
\nabla f(x,y)=
\begin{bmatrix}
2x+y+3\\
x+4y-1
\end{bmatrix},
$$
and the Hessian is
$$
\nabla^2f(x,y)=
\begin{bmatrix}
2&1\\
1&4
\end{bmatrix}.
$$

The characteristic polynomial is
$$
\det\left(\begin{bmatrix}2&1\\1&4\end{bmatrix}-\lambda I\right)
=(2-\lambda)(4-\lambda)-1
=\lambda^2-6\lambda+7.
$$
Thus the eigenvalues are
$$
\lambda=3\pm\sqrt{2},
$$
which are both positive. Therefore the Hessian is positive definite, the objective is strictly convex, and the stationary point is the unique global minimizer.

Solving $\nabla f(x,y)=0$ gives
$$
2x+y+3=0,
\qquad
x+4y-1=0,
$$
so
$$
x^\star=-\frac{13}{7},
\qquad
y^\star=\frac{5}{7}.
$$


```python
import sympy as sp

x, y = sp.symbols("x y")
objective = x ** 2 + x * y + 2 * y ** 2 + 3 * x - y
variables = sp.Matrix([x, y])
gradient = sp.Matrix([sp.diff(objective, variable) for variable in variables])
hessian = sp.hessian(objective, variables)
eigenvalues = hessian.eigenvals()
optimizer_solution = sp.solve(gradient, (x, y), dict=True)[0]
minimum_value = sp.simplify(objective.subs(optimizer_solution))

print("gradient ="); sp.pprint(gradient)
print("Hessian ="); sp.pprint(hessian)
print("eigenvalues =", eigenvalues)
print("optimizer =", optimizer_solution)
print("minimum_value =", minimum_value)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

variables = ca.SX.sym("x", 2)
objective = variables[0] ** 2 + variables[0] * variables[1] + 2 * variables[1] ** 2 + 3 * variables[0] - variables[1]
hessian, gradient = ca.hessian(objective, variables)
linear_term = ca.DM([3, -1])
optimizer = -ca.solve(ca.DM([[2, 1], [1, 4]]), linear_term)
minimum_value = ca.Function("f", [variables], [objective])(optimizer)

print("gradient =", gradient)
print("Hessian =", hessian)
print("optimizer =", optimizer)
print("minimum_value =", minimum_value)
```
