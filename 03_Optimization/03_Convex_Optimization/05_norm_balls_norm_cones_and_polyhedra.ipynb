### 3.3.5. Norm Balls, Norm Cones, and Polyhedra

$$\{x\mid\|x\|\leq r\},\qquad\{(x,t)\mid\|x\|\leq t\}$$


**Explanation:**

[Norm](../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb) balls, norm [cones](03_cones_and_conic_combinations.ipynb), and polyhedra describe convex constraints through magnitudes and affine inequalities. A norm ball limits size, a norm cone compares size with a scalar height, and a polyhedron collects finitely many linear inequalities.

The first set contains all points whose [norm](../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb) is at most the radius $r$. The second set contains pairs $(x,t)$ where the norm of $x$ is bounded by the scalar height $t$, which is the standard norm-cone form.

**Properties:**
- Norm balls are convex by the triangle inequality.
- Norm cones contain pairs whose scalar component dominates the norm.
- Second-order cone constraints use this pattern.

**Intuition:**

The figure above shows a cone boundary; the arithmetic lands exactly on that boundary.

<img src="../../Figures/030205_bv_fig_2_10_second_order_cone.jpeg" alt="Boyd and Vandenberghe Figure 2.10: boundary of the second-order cone" style="width: 60%; height: auto;">


**Numerical Example:**

The figure above shows a cone boundary; the arithmetic lands exactly on that boundary.

Let
$$
u=\begin{bmatrix}3\\4\end{bmatrix},\quad t=5.
$$

Compute the norm:
$$
\|u\|_2=\sqrt{3^2+4^2}=\sqrt{9+16}=\sqrt{25}=5.
$$

The cone test is
$$
\|u\|_2\leq t.
$$

Substitution gives
$$
5\leq5,
$$
so the point is feasible and lies on the cone boundary.


```python
import sympy as sp

vector = sp.Matrix([3, 4])
height = sp.Integer(5)
norm_value = sp.sqrt((vector.T * vector)[0])

print("norm_value =", norm_value)
print("norm_value <= height =", norm_value <= height)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

vector = ca.DM([3, 4])
height = 5
norm_value = ca.norm_2(vector)

print("norm_value =", norm_value)
print("norm_value <= height =", bool(norm_value <= height))
```
