### A2.4.2. Dead Code Elimination

> *Dead code elimination removes instructions whose results are never used and code that is unreachable, reducing program size without changing observable behavior.*

**Explanation:**

**Dead Code Elimination (DCE)** removes two kinds of useless code:

1. **Dead instructions** — instructions whose results are never used by any subsequent instruction or return value.
2. **Unreachable code** — basic blocks that cannot be reached from the function entry.

In SSA form, detecting dead instructions is straightforward: if an instruction's use list is empty (no other instruction references its result) and it has no side effects, it is dead.

**LLVM DCE Passes:**

- **DCE** (`dce`) — simple dead instruction elimination.
- **ADCE** (`adce`) — aggressive DCE; starts from known-live instructions (stores, calls, returns) and marks everything they transitively depend on as live. Everything else is dead.
- **GlobalDCE** (`globaldce`) — removes unreferenced global variables and functions.
- **SimplifyCFG** — removes unreachable blocks.

**Example:**

```llvm
; Before DCE:
%a = add i32 %x, 1       ; used
%b = mul i32 %x, 2       ; DEAD — never used
ret i32 %a

; After DCE:
%a = add i32 %x, 1
ret i32 %a
```

````python
instructions = [
    ("%a", "add i32 %x, 1", True),
    ("%b", "mul i32 %x, 2", False),
    ("%c", "sub i32 %a, 3", True),
    ("%d", "shl i32 %x, 4", False),
    ("ret", "ret i32 %c", True),
]

print("Dead Code Elimination:")
print(f"{'Instruction':<24} {'Live?':<8} {'Action'}")
print("-" * 48)
for name, instruction, is_live in instructions:
    status = "KEEP" if is_live else "REMOVE"
    print(f"{name + ' = ' + instruction if name != 'ret' else instruction:<24} {str(is_live):<8} {status}")

live_instructions = [
    (name, instruction)
    for name, instruction, is_live in instructions
    if is_live
]

print("\nAfter DCE:")
for name, instruction in live_instructions:
    label = f"{name} = {instruction}" if name != "ret" else instruction
    print(f"  {label}")
````
