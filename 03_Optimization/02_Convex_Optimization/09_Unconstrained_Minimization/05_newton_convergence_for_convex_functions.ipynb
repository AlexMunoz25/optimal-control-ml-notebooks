### 3.2.9.5. Newton Convergence for Convex Functions


$$f(x^{(k+1)})-p^\star\leq C(f(x^{(k)})-p^\star)^2$$


**Explanation:**

Newton convergence is rapid when the quadratic model accurately represents the local objective. For an exact quadratic, one full Newton step reaches the minimizer. Feasibility is unconstrained here; optimality is reached when the derivative becomes zero.

The left side is the objective gap after one Newton step. The right side is a constant times the square of the previous gap, which is the objective-value version of quadratic local convergence.

**Properties:**
- Quadratics have constant Hessian.
- The Newton step cancels the current gradient.
- Quadratic convergence appears near the minimizer for general smooth convex functions.

**Intuition:**

The example shows the exact one-step behavior on a scalar quadratic.


**Numerical Example:**

Let
$$
f(x)=(x-3)^2.
$$

At $x^{(0)}=0$,
$$
f'(0)=2(0-3)=-6,\qquad f''(0)=2.
$$

The Newton step is
$$
\Delta x=-\frac{f'(0)}{f''(0)}=-\frac{-6}{2}=3.
$$

Thus
$$
x^{(1)}=0+3=3,
$$
which is the minimizer.


```python
import sympy as sp

initial_point = sp.Integer(0)
gradient = 2 * (initial_point - 3)
hessian = sp.Integer(2)
step = -gradient / hessian
next_point = initial_point + step

step, next_point
```
