### 1.3.2.9. Convexity and the Second Derivative Test

$$
f''(x) > 0 \Rightarrow f \text{ concave up (convex)}, \qquad
f''(x) < 0 \Rightarrow f \text{ concave down} ;
$$

$$
f'(c) = 0,\ f''(c) > 0 \Rightarrow \text{local min}, \qquad
f'(c) = 0,\ f''(c) < 0 \Rightarrow \text{local max}.
$$

**Explanation:**

The second derivative measures concavity: where $f'' > 0$ the slope is increasing and the graph bends upward (convex), and where $f'' < 0$ it bends downward. A point where concavity changes is an *inflection point*. The second derivative test classifies a critical point by curvature there — positive curvature makes it a local minimum, negative a local maximum — and is inconclusive when $f''(c) = 0$. Convexity is the property that makes a stationary point a *global* minimum, which is exactly why it is the central [sufficient optimality condition](../../../03_Optimization/02_Convex_Optimization/03_Convex_Functions/04_second_order_convexity_condition.ipynb) in optimization.

**Properties:**
- A twice-differentiable $f$ is convex on $I$ iff $f''(x) \ge 0$ for all $x \in I$.
- For a convex function every local minimum is a global minimum.

**Intuition:**

A graph is concave up where the slope $f'$ is increasing and concave down where the slope is decreasing.

**Numerical Example:**

Reconsider $f(x) = x^3 - 3x$ with critical points $x = \pm 1$ and $f''(x) = 6x$:

$$
f''(-1) = -6 < 0 \Rightarrow \text{local max at } x = -1, \qquad
f''(1) = 6 > 0 \Rightarrow \text{local min at } x = 1.
$$

Concavity changes sign at $f''(x) = 6x = 0$, i.e. $x = 0$, an inflection point where the curve switches from concave down to concave up. This agrees with the first-derivative-test classification but uses curvature instead of sign changes.

```python
import sympy as sp

x = sp.symbols("x")
f = x**3 - 3 * x
first_derivative = sp.diff(f, x)
second_derivative = sp.diff(f, x, 2)

critical_points = sp.solve(first_derivative, x)
classification = {
    point: ("local min" if second_derivative.subs(x, point) > 0 else "local max")
    for point in critical_points
}
inflection_points = sp.solve(second_derivative, x)

print("f''(x) =", second_derivative)
for point in critical_points:
    print(f"x = {point}: f'' = {second_derivative.subs(x, point)} -> {classification[point]}")
print("inflection point(s):", inflection_points)
```

**Equivalent CasADi implementation:**

CasADi's `ca.hessian` returns the second derivative (a $1\times 1$ Hessian here), whose sign at each critical point gives the same classification.

```python
import casadi as ca

x = ca.SX.sym("x")
f = x**3 - 3 * x
second_derivative, _ = ca.hessian(f, x)
second_derivative_function = ca.Function("d2f", [x], [second_derivative])

for critical_point in (-1.0, 1.0):
    curvature = float(second_derivative_function(critical_point))
    label = "local min" if curvature > 0 else "local max"
    print(f"x = {critical_point}: f'' = {curvature} -> {label}")
```
