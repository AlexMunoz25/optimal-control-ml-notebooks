### 3.1.10. Gradient Method Convergence


$$
\lim_{k\to\infty}\nabla f(x^k)=0,
\qquad
f(x^{k+1}) \le f(x^k),
\qquad
g^k = \nabla f(x^k)+e^k.
$$


**Explanation:**

Convergence analysis asks what the entire generated sequence tells us, not merely whether one step decreased the cost. Under suitable assumptions, such as boundedness of the iterates, smoothness of the cost, and sufficiently good descent, limit points of gradient methods are stationary. This conclusion is weaker than saying the method finds a global minimum, but it is the correct local target for general nonconvex unconstrained problems. Bertsekas also discusses termination, spacer steps, and gradient errors because real algorithms often use approximate derivatives or stop before exact stationarity.

For engineering readers, convergence results should be read as contracts between assumptions and conclusions. If the assumptions are violated, the theorem may no longer describe the computation. If they are satisfied, the theorem explains why the method is at least driven toward points satisfying the first-order necessary condition.

**Example:**

If an error term in the gradient is smaller than the true gradient in norm, the perturbed direction can remain a descent direction. The proof figure shows how a quadratic upper estimate controls the actual cost decrease when the gradient is Lipschitz, a concept naturally expressed using [matrix norm](../../01_Foundations/01_Linear_Algebra/03_Matrix/16_matrix_norm_inequality.ipynb)-type bounds. The code cell then shows gradient norms decreasing along a stable gradient iteration.

<img src="../../Figures/030101_berts_fig_1_2_10_convergence_majorization.jpeg" alt="Bertsekas Figure 1.2.10: quadratic majorization used in a convergence proof" style="width: 60%; height: auto;">


```python
import numpy as np

Q = np.array([[3.0, 0.0], [0.0, 1.0]])
current_point = np.array([3.0, -2.0])
stepsize = 0.25
gradient_norms = []

for iteration in range(12):
    gradient = Q @ current_point
    gradient_norms.append(np.linalg.norm(gradient))
    current_point = current_point - stepsize * gradient

gradient_norms[-5:]
```
