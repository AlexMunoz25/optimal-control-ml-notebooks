### 3.2.3.12. Convexity with Generalized Inequalities


$$f(\theta x+(1-\theta)y)\preceq_K\theta f(x)+(1-\theta)f(y)$$


**Explanation:**

[Cone](../02_Convex_Sets/03_cones_and_conic_combinations.ipynb)-valued convexity replaces the ordinary scalar order with a [generalized inequality](../02_Convex_Sets/09_generalized_inequalities_and_proper_cones.ipynb). The difference between the weighted function values is required to lie in a cone, which lets convexity apply to vector and matrix-valued mappings.

The symbol $\preceq_K$ means comparison using the cone $K$. The formula says the function value at a convex combination is below the same convex combination of function values in the [generalized inequality](../02_Convex_Sets/09_generalized_inequalities_and_proper_cones.ipynb) order.

**Properties:**
- The cone defines the meaning of inequality.
- For the nonnegative orthant, every component must be nonnegative.
- This order supports vector optimization and conic duality.

**Intuition:**

The example checks cone membership by inspecting each component of a difference vector.


**Numerical Example:**

Let
$$
x=\begin{bmatrix}1\\2\end{bmatrix},\quad
y=\begin{bmatrix}3\\5\end{bmatrix},\quad K=\mathbf{R}_+^2.
$$

The order test is $x\preceq_K y$ if $y-x\in K$. Compute
$$
y-x=\begin{bmatrix}3-1\\5-2\end{bmatrix}=\begin{bmatrix}2\\3\end{bmatrix}.
$$

Both components are nonnegative, so $x\preceq_K y$.


```python
import sympy as sp

point_x = sp.Matrix([1, 2])
point_y = sp.Matrix([3, 5])
difference = point_y - point_x
nonnegative = [entry >= 0 for entry in difference]

difference, nonnegative
```
