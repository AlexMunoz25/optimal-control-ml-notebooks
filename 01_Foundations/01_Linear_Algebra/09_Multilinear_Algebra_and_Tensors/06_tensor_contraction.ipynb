### 1.1.9.6. Tensor Contraction

$$
D^j{}_l = C^{ij}{}_{il},
\qquad
\operatorname{rank}(D) = \operatorname{rank}(C) - 2,
\qquad
\frac{\partial x^i}{\partial x^k} = \delta^i{}_k
$$

**Explanation:**

A contraction sets one upper index equal to one lower index and sums over that repeated index. The upper-lower pairing is essential: it is what allows the direct and inverse transformation factors to collapse to the identity, just as in [Einstein Summation Convention](../08_Coordinate_Transformations/09_einstein_summation_convention.ipynb). [Matrix Trace](../03_Matrix/17_matrix_trace.ipynb) is the simplest example of tensor contraction.

**Example:**

Let

$$
A=\begin{bmatrix}1&2\\3&4\end{bmatrix},
\qquad
B=\begin{bmatrix}5&6\\7&8\end{bmatrix}.
$$

Contracting the two indices of $A$ gives its trace:

$$
\operatorname{tr}(A)=1+4=5.
$$

For the constructed rank-$4$ tensor, contracting the matching $A$ indices leaves

$$
5B=5\begin{bmatrix}5&6\\7&8\end{bmatrix}
=\begin{bmatrix}25&30\\35&40\end{bmatrix}.
$$


```python
import sympy as sp

matrix_a = sp.Array([[1, 2], [3, 4]])
matrix_b = sp.Array([[5, 6], [7, 8]])
rank_4_tensor = sp.permutedims(sp.tensorproduct(matrix_a, matrix_b), (0, 2, 1, 3))

matrix_a_trace = sp.tensorcontraction(matrix_a, (0, 1))
contracted_rank_2_tensor = sp.tensorcontraction(rank_4_tensor, (0, 2))
expected_rank_2_tensor = matrix_a_trace * matrix_b

print("matrix_a_trace =", matrix_a_trace)
print("contracted_rank_2_tensor =")
sp.pprint(contracted_rank_2_tensor)
print("expected_rank_2_tensor =")
sp.pprint(expected_rank_2_tensor)
```
