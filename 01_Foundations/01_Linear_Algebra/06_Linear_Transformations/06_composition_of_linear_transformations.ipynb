### 1.1.6.6. Composition of Linear Transformations


$$
(S \circ T)(\mathbf{v}) = S(T(\mathbf{v}))
$$

$$
M_{S \circ T} = M_S M_T
$$


**Explanation:**

Composing two linear transformations means applying one and then the other. Savov shows that the abstract composition law becomes ordinary matrix multiplication once both transformations are written in matrix form. The notebook depends on the definition of a [linear transformation](./01_linear_transformation.ipynb) and on the notebook on [matrix representation](./05_matrix_representation.ipynb).

**Example:**

If $T$ projects onto the $xy$-plane and $S$ deletes the third coordinate, then the composite matrix is the product $M_S M_T$.


```python
import sympy as sp

matrix_t = sp.Matrix([[1, 0], [0, 1], [1, 1]])
matrix_s = sp.Matrix([[1, 0, -1], [0, 1, -1]])
composition_matrix = matrix_s * matrix_t

print("composition_matrix =")
sp.pprint(composition_matrix)
```
