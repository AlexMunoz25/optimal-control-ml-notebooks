### 1.4.6.2. Inverse Laplace Transform

$$
f(t) = \mathcal{L}^{-1}\{F(s)\},
\qquad
\frac{P(s)}{Q(s)} = \sum \frac{A_i}{s - r_i} + \cdots \ (\text{partial fractions}) .
$$

**Explanation:**

Recovering $f(t)$ from its transform $F(s)$ is the **inverse Laplace transform**, also a [linear](./01_definition_of_the_laplace_transform.ipynb) operator, read off the same table in reverse. Since the transforms that arise from solving equations are usually rational functions of $s$, the key algebraic tool is **partial fraction decomposition**: split $F(s)$ into simple pieces whose inverses are table entries. Distinct linear factors give exponentials, repeated factors give $t$-multiplied terms, and irreducible quadratics give sines/cosines (often after completing the square).

**Properties:**
- The "cover-up" method quickly finds the numerators over distinct linear factors.
- Inverse transforms are essentially unique for piecewise-continuous functions of exponential order.

**Numerical Example:**

Evaluate $\mathcal{L}^{-1}\!\left\{\dfrac{s^2 + 6s + 9}{(s-1)(s-2)(s+4)}\right\}$.

**Decompose** with unknown constants:

$$
\frac{s^2 + 6s + 9}{(s-1)(s-2)(s+4)} = \frac{A}{s-1} + \frac{B}{s-2} + \frac{C}{s+4} .
$$

**Cover-up method** (evaluate the left side with each factor removed, at its root):

$$
A = \frac{s^2+6s+9}{(s-2)(s+4)}\Big|_{s=1} = \frac{16}{(-1)(5)} = -\frac{16}{5}, \quad
B = \frac{25}{(1)(6)} = \frac{25}{6}, \quad
C = \frac{1}{(-5)(-6)} = \frac{1}{30} .
$$

**Invert term by term** using $\mathcal{L}^{-1}\{1/(s-a)\} = e^{at}$:

$$
\mathcal{L}^{-1}\!\left\{\frac{s^2+6s+9}{(s-1)(s-2)(s+4)}\right\} = -\frac{16}{5}e^{t} + \frac{25}{6}e^{2t} + \frac{1}{30}e^{-4t} .
$$

```python
import sympy as sp

t, s = sp.symbols("t s", positive=True)
F = (s**2 + 6 * s + 9) / ((s - 1) * (s - 2) * (s + 4))

partial_fractions = sp.apart(F, s)
inverse = sp.inverse_laplace_transform(F, s, t)

print("partial fractions:", partial_fractions)
print("inverse transform:", inverse)
```
