### 3.1.21. Conjugate Gradient Method


$$
\begin{aligned}
g^k &= Qx^k-b, \\
\alpha^k &= \frac{g^{k\mathsf T}g^k}{d^{k\mathsf T}Qd^k}, \\
x^{k+1} &= x^k+\alpha^k d^k, \\
d^{k+1} &= -g^{k+1}+\beta^k d^k.
\end{aligned}
$$


**Explanation:**

The conjugate gradient method generates conjugate directions without first constructing a full conjugate basis. For a [positive definite](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/17_positive_definite_matrix.ipynb) quadratic, it uses gradients and previous search directions to build the next direction, and in exact arithmetic it terminates in at most $n$ steps. Its major practical advantage is that it needs matrix-vector products rather than a factorization of $Q$, making it attractive for large problems.

The vector $g^k=Qx^k-b$ is the gradient of the quadratic model at $x^k$. The stepsize uses inner products with $Q$, and the next direction combines the negative new gradient with the previous direction so the directions remain conjugate.

Conceptually, conjugate gradient improves on steepest descent by remembering useful directional information. The residuals become orthogonal in the Euclidean [inner product](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/02_inner_product_space.ipynb), while the search directions become conjugate in the $Q$ geometry. This combination explains why the method can be much faster than steepest descent while using far less storage than Newton or quasi-Newton methods.

**Intuition:**

In two dimensions, exact conjugate gradient can solve a positive definite quadratic in two iterations. The code cell demonstrates this finite-termination property on a small quadratic system. The example is intentionally small so the reader can connect the algebraic updates with the geometric claim about conjugate directions.


**Numerical Example:**

Solve the positive definite quadratic system $Qx=b$ with
$$
Q=\begin{bmatrix}4&1\\1&3\end{bmatrix},
\qquad b=\begin{bmatrix}1\\2\end{bmatrix},
\qquad x^0=\begin{bmatrix}0\\0\end{bmatrix}.
$$

The first gradient and direction are
$$
g^0=Qx^0-b=\begin{bmatrix}-1\\-2\end{bmatrix},
\qquad d^0=-g^0=\begin{bmatrix}1\\2\end{bmatrix}.
$$

The first stepsize is
$$
\alpha^0=\frac{g^{0\mathsf T}g^0}{d^{0\mathsf T}Qd^0}
=\frac{1+4}{\begin{bmatrix}1&2\end{bmatrix}\begin{bmatrix}6\\7\end{bmatrix}}
=\frac{5}{20}=0.25.
$$

Then
$$
x^1=x^0+0.25d^0=\begin{bmatrix}0.25\\0.5\end{bmatrix},
\qquad g^1=Qx^1-b=\begin{bmatrix}0.5\\-0.25\end{bmatrix}.
$$

The next coefficient is
$$
\beta^0=\frac{g^{1\mathsf T}g^1}{g^{0\mathsf T}g^0}
=\frac{0.25+0.0625}{5}=0.0625,
$$
so
$$
d^1=-g^1+0.0625d^0=\begin{bmatrix}-0.4375\\0.375\end{bmatrix}.
$$

A second exact line step gives
$$
\alpha^1\approx 0.3636,
\qquad
x^2=x^1+\alpha^1d^1=\begin{bmatrix}0.0909\\0.6364\end{bmatrix},
$$
which satisfies $Qx^2=b$.


```python
import casadi as ca

decision_variable = ca.SX.sym("x", 2)
Q = ca.DM([[4, 1], [1, 3]])
linear_term = ca.DM([1, 2])
objective = 0.5 * ca.bilin(Q, decision_variable, decision_variable) - ca.dot(linear_term, decision_variable)
hessian, gradient_expr = ca.hessian(objective, decision_variable)
gradient_function = ca.Function("grad", [decision_variable], [gradient_expr])
hessian_matrix = ca.Function("hessian_eval", [decision_variable], [hessian])(ca.DM([0, 0]))

current_point = ca.DM([0, 0])
gradient = gradient_function(current_point)
direction = -gradient

print("initial_point =", current_point)
print("initial_gradient =", gradient)

for iteration in range(2):
    stepsize = ca.dot(gradient, gradient) / ca.bilin(hessian_matrix, direction, direction)
    current_point = current_point + stepsize * direction
    next_gradient = gradient_function(current_point)
    beta = ca.dot(next_gradient, next_gradient) / ca.dot(gradient, gradient)
    direction = -next_gradient + beta * direction
    gradient = next_gradient
    print(f"iteration {iteration + 1} stepsize =", stepsize)
    print(f"iteration {iteration + 1} current_point =", current_point)
    print(f"iteration {iteration + 1} beta =", beta)

print("final_point =", current_point)
```
