### 3.1.22. Preconditioned Conjugate Gradient and Nonquadratic Use


$$
H z^k = g^k,
\qquad
d^{k+1}=-z^{k+1}+\beta^k d^k,
\qquad
\kappa\!\left(H^{-1/2}QH^{-1/2}\right).
$$


**Explanation:**

Preconditioning changes the geometry in which conjugate gradient operates. The goal is to replace the original spectrum of $Q$ by a more clustered spectrum for the preconditioned matrix, thereby reducing the relevant [condition number](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/31_condition_number.ipynb). For symmetric [positive definite matrices](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/17_positive_definite_matrix.ipynb), the fractional powers in the bound are interpreted through [eigendecomposition](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/12_eigendecomposition.ipynb). In practice, a good preconditioner approximates the hard part of the Hessian while remaining cheap to apply.

The preconditioner $H$ defines an auxiliary [matrix equation](../../01_Foundations/01_Linear_Algebra/04_Computational_Linear_Algebra/06_matrix_equations.ipynb) $Hz^k=g^k$. The vector $z^k$ replaces the raw gradient in the direction update, and the [condition number](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/31_condition_number.ipynb) expression describes the quadratic after scaling by $H^{-1/2}$.

Conjugate-gradient-like methods are also used for nonquadratic problems. Since a nonquadratic Hessian changes from point to point, the finite-termination property no longer applies directly. The method is then used as a local direction generator, often with line search and periodic restarts to recover useful descent behavior.

**Intuition:**

A diagonal preconditioner is the simplest case: it rescales variables according to the diagonal of the quadratic curvature. This does not solve every ill-conditioned problem, but it can help when the main difficulty comes from variable units or diagonal dominance. The code cell compares the condition number before and after such a diagonal preconditioning step.


**Numerical Example:**

Let
$$
Q=\begin{bmatrix}100&5\\5&1\end{bmatrix},
\qquad H=\operatorname{diag}(Q)=\begin{bmatrix}100&0\\0&1\end{bmatrix}.
$$

Then
$$
H^{-1/2}=\begin{bmatrix}0.1&0\\0&1\end{bmatrix}.
$$

The preconditioned matrix is
$$
H^{-1/2}QH^{-1/2}
=\begin{bmatrix}0.1&0\\0&1\end{bmatrix}
\begin{bmatrix}100&5\\5&1\end{bmatrix}
\begin{bmatrix}0.1&0\\0&1\end{bmatrix}
=\begin{bmatrix}1&0.5\\0.5&1\end{bmatrix}.
$$

The [eigenvalues](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/08_eigenvalue.ipynb) of this matrix are
$$
1-0.5=0.5,
\qquad
1+0.5=1.5,
$$
so its condition number is
$$
\frac{1.5}{0.5}=3.
$$

The original matrix has condition number about $134$, so the diagonal preconditioner greatly improves this small example.


```python
import numpy as np

Q = np.array([[100.0, 5.0], [5.0, 1.0]])
preconditioner = np.diag(np.diag(Q))
preconditioned_matrix = np.linalg.inv(np.sqrt(preconditioner)) @ Q @ np.linalg.inv(np.sqrt(preconditioner))
condition_numbers = (np.linalg.cond(Q), np.linalg.cond(preconditioned_matrix))

condition_numbers
```
