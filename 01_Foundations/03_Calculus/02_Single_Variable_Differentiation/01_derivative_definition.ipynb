### 1.3.2.1. Derivative Definition

$$
f'(a) = \lim_{x \to a} \frac{f(x) - f(a)}{x - a}
= \lim_{h \to 0} \frac{f(a+h) - f(a)}{h}
$$

**Explanation:**

The derivative $f'(a)$ is the [limit](../01_Limits_and_Continuity/01_limit_of_a_function.ipynb) of the difference quotient — the slope of the secant line through $(a, f(a))$ and a nearby point — as that second point slides into $a$. Geometrically it is the slope of the tangent line at $a$; physically it is the instantaneous rate of change of $f$. The two displayed forms are equivalent: the second sets $x = a + h$. This single construction, a limit of slopes, is the object every later differentiation rule computes and the quantity gradient-based optimization and machine learning consume.

**Intuition:**

As the second point approaches the first, the secant lines approach the tangent line, whose slope is the derivative.

**Numerical Example:**

Let $f(x) = 3x^2 - 4x + 1$ and compute $f'(2)$ from the definition. Since $f(2) = 12 - 8 + 1 = 5$,

$$
f'(2) = \lim_{x \to 2} \frac{(3x^2 - 4x + 1) - 5}{x - 2}
= \lim_{x \to 2} \frac{3x^2 - 4x - 4}{x - 2}.
$$

Factoring the numerator as $(x - 2)(3x + 2)$ and cancelling,

$$
f'(2) = \lim_{x \to 2} (3x + 2) = 3(2) + 2 = 8.
$$

The tangent line at $x = 2$ therefore has slope $8$.

```python
import sympy as sp

x, h = sp.symbols("x h")
f = 3 * x**2 - 4 * x + 1
base_point = 2

difference_quotient = (f - f.subs(x, base_point)) / (x - base_point)
simplified_quotient = sp.simplify(difference_quotient)
derivative_at_point = sp.limit(difference_quotient, x, base_point)

increment_quotient = (f.subs(x, base_point + h) - f.subs(x, base_point)) / h
derivative_via_h = sp.limit(sp.expand(increment_quotient), h, 0)

print("difference quotient simplifies to:", simplified_quotient)
print("f'(2) via x->a form :", derivative_at_point)
print("f'(2) via h->0 form :", derivative_via_h)
```

**Equivalent CasADi implementation:**

CasADi differentiates the same expression by building its computational graph and applying `ca.gradient`, then evaluates the slope at $x = 2$.

```python
import casadi as ca

x = ca.SX.sym("x")
f = 3 * x**2 - 4 * x + 1
derivative_expression = ca.gradient(f, x)
derivative_function = ca.Function("df", [x], [derivative_expression])

print("f'(x) =", derivative_expression)
print("f'(2) =", float(derivative_function(2)))
```
