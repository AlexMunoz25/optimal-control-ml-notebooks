### 1.1.7.23. Complex Inner Product


$$
\langle \mathbf{u}, \mathbf{v} \rangle = \mathbf{u}^{\mathsf{H}} \mathbf{v}
$$


**Explanation:**

Savov modifies the dot product over $\mathbb{C}$ by conjugating one argument. This notebook isolates that rule because it changes the algebra of orthogonality and norm in a fundamental way. The conjugation operator itself is defined in the previous notebook on [Hermitian transpose](./22_hermitian_transpose.ipynb).

**Example:**

For complex vectors, the inner product of a vector with itself is always real and nonnegative because conjugation cancels the imaginary sign.


```python
import numpy as np

vector_u = np.array([1 + 1j, 2 - 1j])
vector_v = np.array([2.0, 1j])
inner_product = np.vdot(vector_u, vector_v)

print("inner_product =", inner_product)
print("self_inner_product =", np.vdot(vector_u, vector_u))
```
