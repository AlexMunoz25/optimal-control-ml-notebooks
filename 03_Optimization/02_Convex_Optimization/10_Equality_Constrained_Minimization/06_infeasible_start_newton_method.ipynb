### 3.2.10.6. Infeasible-Start Newton Method


$$r(x,\nu)=\begin{bmatrix}\nabla f(x)+A^T\nu\\Ax-b\end{bmatrix}$$


**Explanation:**

Infeasible-start Newton methods reduce a residual that includes both stationarity error and equality-constraint violation. Feasibility is the equality constraint or its linearized version; optimality is stationarity together with that feasibility condition.

The residual vector stacks two errors. The top block is stationarity error $\nabla f(x)+A^T\nu$, and the bottom block is equality-constraint violation $Ax-b$.

**Properties:**
- Equality constraints introduce unrestricted multipliers.
- KKT systems couple stationarity and feasibility.
- Eliminating constraints can reduce the number of variables.

**Intuition:**

The numerical example keeps the equality constraint visible in every algebraic step.


**Numerical Example:**

Let
$$
A=\begin{bmatrix}1&1\end{bmatrix},\quad b=2,
$$
and evaluate at
$$
x=\begin{bmatrix}0\\0\end{bmatrix},\quad \nu=0.
$$

With $\nabla f(x)=0$, the dual residual is
$$
\nabla f(x)+A^T\nu=\begin{bmatrix}0\\0\end{bmatrix}.
$$

The primal residual is
$$
Ax-b=0+0-2=-2.
$$

The point violates feasibility by $-2$.


```python
import sympy as sp

point = sp.Matrix([0, 0])
multiplier = sp.Integer(0)
constraint_matrix = sp.Matrix([[1, 1]])
right_side = sp.Integer(2)
dual_residual = sp.Matrix([0, 0]) + constraint_matrix.T * multiplier
primal_residual = (constraint_matrix * point)[0] - right_side

print("dual_residual =", dual_residual)
print("primal_residual =", primal_residual)
```
