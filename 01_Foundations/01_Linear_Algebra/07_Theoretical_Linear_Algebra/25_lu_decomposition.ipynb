### 1.1.7.25. LU Decomposition


$$
A = L U
$$


**Explanation:**

The LU decomposition factors a matrix into lower- and upper-triangular parts. Its value is computational: triangular factors make linear systems easier to solve by forward and backward substitution. The positive-definite special case appears in [Cholesky decomposition](./19_cholesky_decomposition.ipynb).

**Example:**

If $A = LU$, then solving $A \mathbf{x} = \mathbf{b}$ becomes two triangular solves: first $L \mathbf{y} = \mathbf{b}$ and then $U \mathbf{x} = \mathbf{y}$.


```python
import sympy as sp

matrix_a = sp.Matrix([[3, 1], [1, 2]])
matrix_l, matrix_u, _ = matrix_a.LUdecomposition()

print("matrix_l =")
sp.pprint(matrix_l)
print("matrix_u =")
sp.pprint(matrix_u)
```
