### 1.4.4.5. Variation of Parameters

$$
y_p = u_1 y_1 + u_2 y_2, \qquad
u_1' = -\frac{y_2 f(x)}{W}, \quad u_2' = \frac{y_1 f(x)}{W}, \quad
W = W(y_1, y_2),
$$

for $y'' + P(x)y' + Q(x)y = f(x)$.

**Explanation:**

Variation of parameters finds a particular solution for **any** continuous forcing $f(x)$, not just the special forms of [undetermined coefficients](./04_undetermined_coefficients.ipynb). Starting from the complementary solution $y_c = c_1 y_1 + c_2 y_2$, it replaces the constants by functions: $y_p = u_1 y_1 + u_2 y_2$. Imposing the natural constraint $y_1 u_1' + y_2 u_2' = 0$ leads, via Cramer's rule, to $u_1'$ and $u_2'$ expressed through the [Wronskian](./01_preliminary_theory_of_linear_equations.ipynb) $W$; integrating gives $u_1, u_2$. It is the one method that always works for linear equations and generalizes to the [variation-of-constants formula](../08_Systems_of_Linear_First_Order_ODEs/04_matrix_exponential.ipynb) for systems.

**Properties:**
- Put the equation in standard form (lead coefficient 1) to read off $f(x)$.
- $W \neq 0$ for a fundamental set, so the formulas for $u_1', u_2'$ are always defined.

**Numerical Example:**

Solve $y'' - 4y' + 4y = (x + 1)e^{2x}$.

**Step 1 — complementary function.** Auxiliary $(m-2)^2 = 0$, so $y_1 = e^{2x}$, $y_2 = xe^{2x}$ and $y_c = c_1 e^{2x} + c_2 x e^{2x}$.

**Step 2 — Wronskian.**

$$
W(e^{2x}, xe^{2x}) =
\begin{vmatrix} e^{2x} & xe^{2x} \\ 2e^{2x} & 2xe^{2x} + e^{2x} \end{vmatrix} = e^{4x}.
$$

**Step 3 — the derivatives.** With $f(x) = (x+1)e^{2x}$,

$$
u_1' = -\frac{y_2 f}{W} = -\frac{xe^{2x}(x+1)e^{2x}}{e^{4x}} = -x^2 - x, \qquad
u_2' = \frac{y_1 f}{W} = \frac{e^{2x}(x+1)e^{2x}}{e^{4x}} = x + 1 .
$$

**Step 4 — integrate.** $u_1 = -\tfrac13 x^3 - \tfrac12 x^2$ and $u_2 = \tfrac12 x^2 + x$, so

$$
y_p = u_1 y_1 + u_2 y_2 = \tfrac16 x^3 e^{2x} + \tfrac12 x^2 e^{2x},
\qquad y = c_1 e^{2x} + c_2 x e^{2x} + \tfrac16 x^3 e^{2x} + \tfrac12 x^2 e^{2x} .
$$

```python
import sympy as sp

x = sp.symbols("x")
y = sp.Function("y")

y_1, y_2 = sp.exp(2 * x), x * sp.exp(2 * x)
forcing = (x + 1) * sp.exp(2 * x)
wronskian = sp.Matrix([[y_1, y_2], [y_1.diff(x), y_2.diff(x)]]).det()

u_1 = sp.integrate(-y_2 * forcing / wronskian, x)
u_2 = sp.integrate(y_1 * forcing / wronskian, x)
particular = sp.expand(u_1 * y_1 + u_2 * y_2)
general = sp.dsolve(sp.Eq(y(x).diff(x, 2) - 4 * y(x).diff(x) + 4 * y(x), forcing), y(x))

print("W(y1, y2) =", sp.simplify(wronskian))
print("u1' integrates to u1 =", u_1, "   u2' integrates to u2 =", u_2)
print("y_p =", particular)
print("general solution:", sp.expand(general.rhs))
```
