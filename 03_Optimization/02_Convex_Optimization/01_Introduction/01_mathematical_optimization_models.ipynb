### 3.2.1.1. Mathematical Optimization Models


$$\min_{x} f_0(x)\quad\text{subject to}\quad f_i(x)\leq0,\ h_i(x)=0$$


**Explanation:**

An optimization model separates the decision variable, feasible set, and objective value. The feasible set determines which candidates may be compared, and the objective ranks only those candidates. In convex optimization, this separation lets first-order conditions, geometric arguments, and dual bounds certify global optimality.

The display reads as a constrained minimization problem. The [vector](../../../01_Foundations/01_Linear_Algebra/02_Vector/01_vector_definition.ipynb) $x$ is the decision variable, $f_0(x)$ is the objective being minimized, $f_i(x)\leq0$ are inequality constraints, and $h_i(x)=0$ are equality constraints.

**Properties:**
- Feasibility is checked before objective values are compared.
- Optimality means no feasible point has a smaller objective value.
- Convex structure makes global certificates possible.

**Intuition:**

Think of the constraints as the search region and the objective as the ruler used inside that region.


**Numerical Example:**

Let $X=\{1,2,3\}$ and $f(x)=(x-2)^2$. First check that all three candidates are feasible because each belongs to $X$.

Then
$$
f(1)=(1-2)^2=1,
$$
$$
f(2)=(2-2)^2=0,
$$
and
$$
f(3)=(3-2)^2=1.
$$

The smallest feasible value is $0$, so the optimizer is $x^\star=2$.


```python
import sympy as sp

candidates = [sp.Integer(1), sp.Integer(2), sp.Integer(3)]
values = [(candidate - 2) ** 2 for candidate in candidates]
optimizer = candidates[values.index(min(values))]

print("optimizer =", optimizer)
print("values =", values)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

opti = ca.Opti()
decision_variable = opti.variable()
opti.minimize((decision_variable - 2) ** 2)
opti.solver("ipopt", {"print_time": False}, {"print_level": 0, "sb": "yes"})
solution = opti.solve()

print("optimizer =", ca.DM(solution.value(decision_variable)))
```
