### 1.3.6.1. Double and Iterated Integrals

$$
\iint_{R} f(x, y)\,dA = \int_{a}^{b}\!\int_{c}^{d} f(x, y)\,dy\,dx
= \int_{c}^{d}\!\int_{a}^{b} f(x, y)\,dx\,dy \qquad (\text{Fubini}) .
$$

**Explanation:**

A double integral accumulates a function over a two-dimensional region, generalizing the [definite integral](../03_Single_Variable_Integration/02_definite_integral_and_riemann_sums.ipynb) as a limit of sums over area elements $dA$. Fubini's theorem evaluates it as an *iterated* integral: integrate with respect to one variable while holding the other fixed (the integral analogue of a [partial derivative](../05_Multivariable_Differential_Calculus/02_partial_derivatives.ipynb)), then integrate the result. For a continuous integrand the order of integration may be swapped. Double and iterated integrals are how multivariable expectations and total quantities over a domain are computed in probability and modelling.

**Properties:**
- Over a rectangle the limits are constants; over a general region the inner limits may depend on the outer variable.
- Linearity and the order-swap (Fubini) hold whenever $f$ is integrable over $R$.

**Intuition:**

The double integral measures the volume between the surface $z = f(x,y)$ and the rectangular region beneath it.

**Numerical Example:**

Evaluate $\iint_{R} xy\,dA$ over $R = [0,1]\times[0,2]$. Integrate first in $y$ (treating $x$ as constant):

$$
\int_{0}^{2} xy\,dy = x\left[\frac{y^2}{2}\right]_{0}^{2} = x\cdot 2 = 2x .
$$

Then integrate in $x$:

$$
\int_{0}^{1} 2x\,dx = \left[x^2\right]_{0}^{1} = 1 .
$$

So $\iint_R xy\,dA = 1$. Reversing the order (first $x$, then $y$) gives $\int_0^2 \tfrac{y}{2}\,dy = 1$, confirming Fubini.

```python
import sympy as sp

x, y = sp.symbols("x y")
f = x * y

inner_in_y = sp.integrate(f, (y, 0, 2))
iterated_value = sp.integrate(inner_in_y, (x, 0, 1))
reversed_order = sp.integrate(sp.integrate(f, (x, 0, 1)), (y, 0, 2))

print("∫_0^2 xy dy            =", inner_in_y)
print("∫_0^1 (2x) dx          =", iterated_value)
print("reversed order (x then y) =", reversed_order)
```
