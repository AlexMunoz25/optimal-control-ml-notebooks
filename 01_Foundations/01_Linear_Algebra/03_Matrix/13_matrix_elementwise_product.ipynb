### 1.1.3.13. Matrix Element-wise Product (Hadamard)

$$
(\mathbf{A} \circ \mathbf{B})_{ij} = A_{ij} B_{ij}
$$

**Explanation:**

The Hadamard product multiplies matrices element by element.
Given two matrices $\mathbf{A}$ and $\mathbf{B}$ of the same shape, the element-wise product $\mathbf{A} \circ \mathbf{B}$ is defined entry-wise.
Different from standard matrix multiplication which uses dot products of rows and columns.

Compare this entry-wise operation with standard [Matrix Multiplication](./05_matrix_multiplication.ipynb), which uses row-column products. It uses the same entry indexing introduced in [Matrices](./01_matrix_definition.ipynb).

**Example:**

Let

$$
\mathbf{A} =
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix},
\quad
\mathbf{B} =
\begin{bmatrix}
5 & 6 \\
7 & 8
\end{bmatrix}.
$$

The Hadamard product is

$$
\mathbf{A} \circ \mathbf{B} =
\begin{bmatrix}
5 & 12 \\
21 & 32
\end{bmatrix}.
$$


```python
import sympy as sp

matrix_a = sp.Matrix([[1, 2], [3, 4]])
matrix_b = sp.Matrix([[5, 6], [7, 8]])

hadamard_product = matrix_a.multiply_elementwise(matrix_b)

print("A ∘ B =")
sp.pprint(hadamard_product)
```
