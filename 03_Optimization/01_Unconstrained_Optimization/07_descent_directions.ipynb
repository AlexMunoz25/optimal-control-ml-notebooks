### 3.1.7. Descent Directions


$$
x^{k+1} = x^k + \alpha^k d^k,
\qquad
\nabla f(x^k)^{\mathsf T}d^k < 0,
\qquad
\alpha^k > 0.
$$


**Explanation:**

A descent direction is a direction in which the cost initially decreases. The inequality $\nabla f(x^k)^{\mathsf T}d^k<0$ says that the directional derivative at the current point is negative, so Taylor's formula guarantees a decrease for all sufficiently small positive stepsizes. Geometrically, the direction makes an obtuse angle with the gradient, where the angle is understood through the usual [inner product](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/02_inner_product_space.ipynb). This is the local mechanism behind most algorithms in Chapter 1.

It is useful to separate the direction from the stepsize. The direction decides where the method tries to move; the stepsize decides how far it actually moves along that direction. Bertsekas builds gradient methods from exactly these two choices because different algorithms mainly differ in how they choose $d^k$ and $\alpha^k$.

```python
import numpy as np

Q = np.array([[3.0, 0.0], [0.0, 1.0]])
current_point = np.array([1.0, 2.0])
gradient = Q @ current_point
direction = -gradient
stepsizes = np.array([0.05, 0.2, 0.8])
objective_values = [0.5 * (current_point + stepsize * direction) @ Q @ (current_point + stepsize * direction) for stepsize in stepsizes]

0.5 * current_point @ Q @ current_point, objective_values
```
