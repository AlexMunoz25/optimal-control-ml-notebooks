### 1.1.8.9. Einstein Summation Convention


$$
A^{i'} = \sum_{j=1}^{3} a_{ij} A^j = a_{ij} A^j,
\qquad
A_{i'} = \sum_{j=1}^{3} b_{ij} A_j = b_{ij} A_j
$$


**Explanation:**

Einstein summation convention suppresses the summation sign when an index appears twice in one term, once up and once down. The repeated index is a dummy index and is summed over, while the free index labels the output component. This notation turns long component-wise transformation rules into compact expressions without changing the underlying algebra.

**Example:**

The three equations
$$
\begin{aligned}
A^{1'} &= a_{11} A^1 + a_{12} A^2 + a_{13} A^3, \\
A^{2'} &= a_{21} A^1 + a_{22} A^2 + a_{23} A^3, \\
A^{3'} &= a_{31} A^1 + a_{32} A^2 + a_{33} A^3
\end{aligned}
$$
are written compactly as
$$
A^{i'} = a_{ij} A^j.
$$


```python
import numpy as np

transformation_matrix = np.array(
    [
        [2.0, -1.0, 0.0],
        [1.0, 3.0, 4.0],
        [0.0, 2.0, 1.0],
    ]
)
contravariant_components = np.array([1.0, 2.0, -1.0])
transformed_components_from_matrix_product = transformation_matrix @ contravariant_components
transformed_components_from_explicit_sums = np.array(
    [sum(matrix_row * contravariant_components) for matrix_row in transformation_matrix]
)

print("transformed_components_from_matrix_product =", transformed_components_from_matrix_product)
print("transformed_components_from_explicit_sums =", transformed_components_from_explicit_sums)
```
