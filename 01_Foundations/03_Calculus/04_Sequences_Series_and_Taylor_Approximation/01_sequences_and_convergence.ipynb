### 1.3.4.1. Sequences and Convergence

$$
\lim_{n \to \infty} a_n = L \iff
\forall \varepsilon > 0\ \exists N : n \ge N \Rightarrow |a_n - L| < \varepsilon .
$$

**Explanation:**

A sequence is an ordered list $a_1, a_2, a_3, \ldots$ — a function from the positive integers to the reals. It *converges* to a limit $L$ if its terms get and stay arbitrarily close to $L$ for all sufficiently large $n$; otherwise it diverges. This is the discrete version of a [limit of a function](../01_Limits_and_Continuity/01_limit_of_a_function.ipynb), and it is the exact language used to say that an iterative algorithm — a [gradient method](../../../03_Optimization/01_Unconstrained_Optimization/10_gradient_method_convergence.ipynb) producing iterates $x_1, x_2, \ldots$ — converges to a solution.

**Properties:**
- A monotone bounded sequence converges (monotone convergence theorem).
- The limit laws carry over: limits of sums, products, and quotients of convergent sequences behave as expected.

**Numerical Example:**

Consider $a_n = \dfrac{2n + 1}{n}$. Dividing through by $n$,

$$
a_n = 2 + \frac{1}{n} \;\longrightarrow\; 2 \quad\text{as } n \to \infty,
$$

since $1/n \to 0$. Numerically $a_1 = 3$, $a_{10} = 2.1$, $a_{100} = 2.01$, $a_{1000} = 2.001$ — the terms close in on $L = 2$. By contrast $b_n = (-1)^n$ oscillates between $-1$ and $1$ and has no limit, so it diverges.

```python
import sympy as sp

n = sp.symbols("n", positive=True, integer=True)
convergent_sequence = (2 * n + 1) / n

sequence_limit = sp.limit(convergent_sequence, n, sp.oo)
sampled_terms = [(value, sp.N(convergent_sequence.subs(n, value), 6)) for value in [1, 10, 100, 1000]]

print("a_n = (2n+1)/n -> limit:", sequence_limit)
for index, term in sampled_terms:
    print(f"a_{index:<4} = {term}")
```
