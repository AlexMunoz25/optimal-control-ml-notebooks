### A2.3.2. Analysis Pass

> *An analysis pass computes information about the IR without modifying it. Its results are cached by the analysis manager and invalidated only when a transformation pass reports that the analysis may no longer be valid.*

**Explanation:**

An **analysis pass** inspects the IR and produces a result that transformation passes consume. Analyses are *lazy* — they are computed only when requested and cached for reuse.

**Structure of an Analysis:**

1. **Result type** — a struct or class holding the computed information.
2. **`run` method** — takes the IR unit and the analysis manager, returns the result.
3. **Invalidation** — the result is discarded when a transformation pass declares it is no longer preserved.

**Built-in Analyses:**

- `DominatorTreeAnalysis` — computes the dominator tree of a function.
- `LoopAnalysis` — identifies natural loops.
- `AAManager` (AliasAnalysis) — determines whether two pointers may alias.
- `ScalarEvolutionAnalysis` — models loop induction variables.

**Example:**

```cpp
struct BlockCountAnalysis : public AnalysisInfoMixin<BlockCountAnalysis> {
  struct Result {
    unsigned count;
  };
  Result run(Function &F, FunctionAnalysisManager &) {
    return {static_cast<unsigned>(F.size())};
  }
  static AnalysisKey Key;
};
```

````python
builtin_analyses = [
    ("DominatorTreeAnalysis", "Dominator tree", "Loop detection, SSA construction"),
    ("LoopAnalysis", "Natural loop nesting", "Loop optimizations (LICM, unroll)"),
    ("AAManager", "Alias analysis", "Load/store reordering, dead store elimination"),
    ("ScalarEvolutionAnalysis", "Induction variable modeling", "Loop bound analysis, trip count"),
    ("MemorySSAAnalysis", "Memory SSA form", "Memory-aware optimizations"),
]

print("Key LLVM Analyses:")
print(f"{'Analysis':<28} {'Computes':<28} {'Used By'}")
print("-" * 84)
for name, computes, used_by in builtin_analyses:
    print(f"{name:<28} {computes:<28} {used_by}")
````
