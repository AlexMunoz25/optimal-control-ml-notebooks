### 1.4.2.3. Separable Equations

$$
\frac{dy}{dx} = g(x)\,h(y) \;\Longrightarrow\; \int \frac{dy}{h(y)} = \int g(x)\,dx .
$$

**Explanation:**

A first-order equation is **separable** when its right side factors as a function of $x$ times a function of $y$. Then the variables can be moved to opposite sides and each side [integrated](../../03_Calculus/03_Single_Variable_Integration/01_antiderivative_and_indefinite_integral.ipynb) independently — often after a [substitution](../../03_Calculus/03_Single_Variable_Integration/04_integration_by_substitution.ipynb) — producing a one-parameter family of solutions, usually in implicit form. Only one constant of integration is needed (the two combine). Care is required where $h(y) = 0$: those constant solutions may be lost in the division and must be checked separately.

**Numerical Example:**

**Example A** — solve $(1 + x)\,dy - y\,dx = 0$. Divide by $(1+x)\,y$ to separate:

$$
\frac{dy}{y} = \frac{dx}{1 + x}.
$$

Integrate both sides:

$$
\int \frac{dy}{y} = \int \frac{dx}{1 + x} \;\Rightarrow\; \ln|y| = \ln|1 + x| + c_1 .
$$

Exponentiating and writing $\pm e^{c_1} = c$,

$$
|y| = e^{c_1}|1 + x| \;\Rightarrow\; y = c\,(1 + x).
$$

**Example B (IVP)** — solve $\dfrac{dy}{dx} = -\dfrac{x}{y}$, $y(4) = -3$. Separate as $y\,dy = -x\,dx$ and integrate:

$$
\int y\,dy = -\int x\,dx \;\Rightarrow\; \frac{y^2}{2} = -\frac{x^2}{2} + c_1 \;\Rightarrow\; x^2 + y^2 = c^2 .
$$

Apply the initial condition $x = 4,\ y = -3$: $\;16 + 9 = 25 = c^2$. The implicit solution is the circle $x^2 + y^2 = 25$; the branch through $(4, -3)$ is the lower semicircle

$$
y = -\sqrt{25 - x^2}, \qquad -5 < x < 5 .
$$

```python
import sympy as sp

x = sp.symbols("x")
y = sp.Function("y")

example_a = sp.dsolve(sp.Eq((1 + x) * y(x).diff(x), y(x)), y(x))
example_b = sp.dsolve(sp.Eq(y(x).diff(x), -x / y(x)), y(x), ics={y(4): -3})

print("Example A: y =", example_a.rhs)
print("Example B (IVP): y =", example_b.rhs)
print("check B at x=4 :", sp.simplify(example_b.rhs.subs(x, 4)))
```
