### 1.1.7.24. Unitary Matrix


$$
U^{\mathsf{H}} U = I
$$


**Explanation:**

A unitary matrix is the complex analogue of an [orthogonal matrix](./11_orthogonal_matrix.ipynb). It preserves the complex inner product and therefore preserves lengths and angles in complex vector spaces.

**Example:**

A diagonal matrix with entries of unit magnitude is unitary because each diagonal entry cancels with its complex conjugate in the Hermitian product.


```python
import sympy as sp

matrix_u = sp.Matrix([[1, 0], [0, sp.I]])

print("unitary_check =")
sp.pprint(matrix_u.conjugate().T * matrix_u)
```
