### 3.9.23. Quasi-Newton Scaling and Comparison

$$
\tilde D^0
= \frac{p^{0\mathsf T}q^0}{q^{0\mathsf T}D^0q^0}D^0,
\qquad
R^k = Q^{1/2}D^kQ^{1/2}.
$$


**Explanation:**

The initial scaling of a quasi-Newton matrix can strongly influence the early iterations, before the method has accumulated enough curvature pairs. Scaling matters because multiplying $D^0$ by a suitable positive scalar can make the first approximate inverse Hessian better match the observed curvature. The [condition number](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/31_condition_number.ipynb) of $R^k$ describes the scaled local geometry, connecting quasi-Newton scaling back to [spectral properties](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/30_spectral_radius.ipynb), [eigendecomposition](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/12_eigendecomposition.ipynb), and [condition numbers](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/31_condition_number.ipynb).

The first formula rescales the initial inverse-Hessian approximation using the first step $p^0$ and gradient change $q^0$. The second formula forms a similarity-scaled matrix $R^k$, so its [condition number](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/31_condition_number.ipynb) describes the local geometry of the quasi-Newton iteration.

The comparison with conjugate gradient is also practical. Quasi-Newton methods tend to be strong when gradient evaluations are expensive enough that the extra matrix storage and [matrix-vector multiplication](../../01_Foundations/01_Linear_Algebra/03_Matrix/04_matrix_vector_multiplication.ipynb) are acceptable. Conjugate gradient is often preferred in very large problems where storing a full matrix is too costly and matrix-vector products are cheap.

**Intuition:**

A scalar rescaling of $D^0$ uses the first step and gradient change to match curvature along the observed direction. The figure relates this scaling idea to the condition number of a quasi-Newton iteration. The code cell computes the simple scalar scaling factor from one curvature pair.


**Numerical Example:**

The condition-number figure motivates scaling the initial inverse-Hessian approximation. Let
$$
D^0=I,
\qquad p^0=\begin{bmatrix}1\\2\end{bmatrix},
\qquad q^0=\begin{bmatrix}3\\5\end{bmatrix}.
$$

The numerator is
$$
p^{0\mathsf T}q^0=1\cdot 3+2\cdot 5=13.
$$

The denominator is
$$
q^{0\mathsf T}D^0q^0=q^{0\mathsf T}q^0=3^2+5^2=34.
$$

Therefore the scalar scaling factor is
$$
\frac{p^{0\mathsf T}q^0}{q^{0\mathsf T}D^0q^0}=\frac{13}{34}\approx 0.3824.
$$

The scaled starting matrix is
$$
\tilde{D}^0=0.3824I
=\begin{bmatrix}0.3824&0\\0&0.3824\end{bmatrix}.
$$


```python
import sympy as sp

initial_inverse_hessian = sp.eye(2)
position_increment = sp.Matrix([1, 2])
gradient_increment = sp.Matrix([3, 5])
scale_factor = (position_increment.T * gradient_increment)[0] / (gradient_increment.T * initial_inverse_hessian * gradient_increment)[0]
scaled_inverse_hessian = scale_factor * initial_inverse_hessian

print("scale_factor =", scale_factor)
print("scaled_inverse_hessian ="); sp.pprint(scaled_inverse_hessian)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

initial_inverse_hessian = ca.DM.eye(2)
position_increment = ca.DM([1, 2])
gradient_increment = ca.DM([3, 5])
scale_factor = ca.dot(position_increment, gradient_increment) / ca.bilin(initial_inverse_hessian, gradient_increment, gradient_increment)
scaled_inverse_hessian = scale_factor * initial_inverse_hessian

print("scale_factor =", scale_factor)
print("scaled_inverse_hessian =", scaled_inverse_hessian)
```
