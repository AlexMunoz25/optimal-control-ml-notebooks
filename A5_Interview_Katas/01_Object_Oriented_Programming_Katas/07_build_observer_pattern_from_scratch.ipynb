### A5.1.7. Build Observer Pattern from Scratch

**Problem:**

Build a subject class that maintains a list of observers. When the subject's state changes, all registered observers are notified automatically.

**Explanation:**

The observer pattern lets multiple objects watch a subject for changes. The subject keeps a list of callbacks or observer objects. Whenever something interesting happens, the subject iterates the list and calls each observer.

How to build it:

1. Create a `Subject` class with a list of observers in `__init__`.
2. `subscribe(observer)`: append the observer to the list.
3. `unsubscribe(observer)`: remove the observer from the list.
4. `notify(event)`: iterate all observers and call their `update(event)` method.
5. Observers are any object with an `update` method.

**Example:**

A `TemperatureSensor` subject notifies a `Display` and a `Logger` whenever the temperature changes. Both receive the new value without the sensor knowing their concrete types.

```python
class Subject:
    def __init__(self):
        self.observers = []

    def subscribe(self, observer):
        self.observers.append(observer)

    def unsubscribe(self, observer):
        self.observers.remove(observer)

    def notify(self, event):
        for observer in self.observers:
            observer.update(event)


class TemperatureSensor(Subject):
    def __init__(self):
        super().__init__()
        self.temperature = 0.0

    def set_temperature(self, value):
        self.temperature = value
        self.notify(value)


class Display:
    def update(self, temperature):
        print(f"Display: {temperature}°C")


class Logger:
    def __init__(self):
        self.log = []

    def update(self, temperature):
        self.log.append(temperature)
        print(f"Logger recorded: {temperature}°C (total entries: {len(self.log)})")


sensor = TemperatureSensor()
display = Display()
logger = Logger()

sensor.subscribe(display)
sensor.subscribe(logger)

sensor.set_temperature(22.5)
sensor.set_temperature(23.1)
```
