### 3.3.11. Constraint Qualifications: MFCQ and Slater

$$
\nabla h_i(x^*)^T d=0,
\qquad
\nabla g_j(x^*)^T d<0,\quad j\in A(x^*)
$$

$$
g_j(\bar{x})<0,\quad j\in A(x^*)
$$

**Explanation:**

Constraint qualifications are conditions that rule out abnormal Fritz John multipliers and recover KKT multipliers. The Mangasarian-Fromovitz condition asks for a direction that preserves equality constraints to first order while moving strictly inside every active inequality. Slater's condition is a convex-inequality version: a strictly feasible point supplies the needed inward direction.

**Intuition:**

<img src="../../Figures/03030311_berts_fig_3_3_2_mfcq.jpeg" alt="Bertsekas Figure 3.3.2: a common inward direction for active inequality gradients in the Mangasarian-Fromovitz condition" style="width: 60%; height: auto;">

The figure illustrates the role of $d$: all active inequality gradients have negative inner product with the same direction, so no nonzero nonnegative combination of active gradients can cancel the objective-gradient multiplier abnormally.

**Numerical Example:**

Consider the active constraints at

$$
x^*=\begin{bmatrix}0\\0\end{bmatrix}
$$

with

$$
g_1(x)=x_1\le 0,
\qquad
g_2(x)=x_2\le 0.
$$

Both constraints are active because

$$
g_1(x^*)=0, \qquad g_2(x^*)=0.
$$

The active gradients are

$$
\nabla g_1(x^*)=\begin{bmatrix}1\\0\end{bmatrix},
\qquad
\nabla g_2(x^*)=\begin{bmatrix}0\\1\end{bmatrix}.
$$

Choose

$$
d=\begin{bmatrix}-1\\-1\end{bmatrix}.
$$

Then

$$
\nabla g_1(x^*)^T d = -1 <0,
\qquad
\nabla g_2(x^*)^T d = -1 <0.
$$

Thus MFCQ holds. For the convex Slater condition, the point

$$
\bar{x}=\begin{bmatrix}-1\\-1\end{bmatrix}
$$

satisfies

$$
g_1(\bar{x})=-1<0,
\qquad
g_2(\bar{x})=-1<0.
$$

The same geometry verifies both the inward-direction view and the strict-feasibility view.

```python
import sympy as sp

gradient_constraint_1 = sp.Matrix([1, 0])
gradient_constraint_2 = sp.Matrix([0, 1])
inward_direction = sp.Matrix([-1, -1])
slater_point = sp.Matrix([-1, -1])

mfcq_products = sp.Matrix([
    gradient_constraint_1.dot(inward_direction),
    gradient_constraint_2.dot(inward_direction),
])
slater_values = sp.Matrix([
    slater_point[0],
    slater_point[1],
])

result = {
    'mfcq_products': mfcq_products,
    'slater_values': slater_values,
}
result
```
