### 3.1.7. Descent Directions


$$
x^{k+1} = x^k + \alpha^k d^k,
\qquad
\nabla f(x^k)^{\mathsf T}d^k < 0,
\qquad
\alpha^k > 0.
$$


**Explanation:**

A descent direction is a direction in which the cost initially decreases. The inequality $\nabla f(x^k)^{\mathsf T}d^k<0$ says that the directional derivative at the current point is negative, so Taylor's formula guarantees a decrease for all sufficiently small positive stepsizes. Geometrically, the direction makes an obtuse angle with the gradient, where the angle is understood through the usual [inner product](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/02_inner_product_space.ipynb). This is the local mechanism behind many unconstrained optimization algorithms.

It is useful to separate the direction from the stepsize. The direction decides where the method tries to move; the stepsize decides how far it actually moves along that direction. Gradient methods are built from exactly these two choices because different algorithms mainly differ in how they choose $d^k$ and $\alpha^k$.

**Intuition:**

The first figure introduces iterative descent as a sequence of points with decreasing cost. The next two figures show why the negative gradient and more general directions with negative directional derivative can reduce the objective, at least for small steps. The code cell evaluates the same small negative-gradient step used in the numerical example.

**Numerical Example:**

The descent-direction figures show that a small step opposite the gradient lowers the cost. Let
$$
Q=\begin{bmatrix}3&0\\0&1\end{bmatrix},\qquad f(x)=\frac{1}{2}x^{\mathsf T}Qx,\qquad x^k=\begin{bmatrix}1\\2\end{bmatrix}.
$$

The gradient and negative-gradient direction are
$$
\nabla f(x^k)=Qx^k=\begin{bmatrix}3\\2\end{bmatrix},
\qquad d^k=-\nabla f(x^k)=\begin{bmatrix}-3\\-2\end{bmatrix}.
$$

The directional derivative is
$$
\nabla f(x^k)^{\mathsf T}d^k=3(-3)+2(-2)=-9-4=-13<0,
$$
so $d^k$ is a descent direction. With $\alpha=0.2$,
$$
x^k+\alpha d^k=\begin{bmatrix}1\\2\end{bmatrix}+0.2\begin{bmatrix}-3\\-2\end{bmatrix}
=\begin{bmatrix}0.4\\1.6\end{bmatrix}.
$$

The cost decreases from
$$
f(x^k)=\frac{1}{2}(3\cdot 1^2+2^2)=\frac{7}{2}=3.5
$$
to
$$
f(x^k+\alpha d^k)=\frac{1}{2}(3\cdot 0.4^2+1.6^2)=\frac{1}{2}(0.48+2.56)=1.52.
$$


```python
import numpy as np

Q = np.array([[3.0, 0.0], [0.0, 1.0]])
current_point = np.array([1.0, 2.0])
gradient = Q @ current_point
direction = -gradient
stepsize = 0.2
current_objective = 0.5 * current_point @ Q @ current_point
next_point = current_point + stepsize * direction
next_objective = 0.5 * next_point @ Q @ next_point

current_objective, next_point, next_objective
```
