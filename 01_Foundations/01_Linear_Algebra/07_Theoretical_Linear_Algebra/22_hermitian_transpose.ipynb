### 1.1.7.22. Hermitian Transpose


$$
A^{\mathsf{H}} = \overline{A}^{\mathsf{T}}
$$


**Explanation:**

The Hermitian transpose is the complex analogue of the ordinary transpose, obtained by transposing and conjugating. It is the operator that makes orthogonality and inner products over $\mathbb{C}$ work correctly, and it leads directly to [complex inner products](./23_complex_inner_product.ipynb).

**Example:**

If

$$
A =
\begin{bmatrix}
1 + i & 2 \\
3i & 4 - i
\end{bmatrix},
$$

then each entry is conjugated and reflected across the diagonal.


```python
import sympy as sp

matrix_a = sp.Matrix([[1 + sp.I, 2], [3 * sp.I, 4 - sp.I]])

print("hermitian_transpose =")
sp.pprint(matrix_a.conjugate().T)
```
