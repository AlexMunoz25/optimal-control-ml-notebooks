### 3.2.11.2. Logarithmic Barrier Function


$$\phi(x)=-\sum_i\log(-f_i(x))$$


**Explanation:**

The logarithmic barrier turns each strict inequality into a smooth penalty term. It is finite inside the feasible region and grows without bound near the boundary, so unconstrained minimization of the barrier objective favors interior points.

**Properties:**
- The log barrier requires every inequality value to be negative.
- The barrier is finite in the strict interior.
- Interior-point methods maintain strict feasibility while reducing the approximation gap.

**Intuition:**

The figure above shows barrier growth near a boundary; the arithmetic evaluates one interior point.

<img src="../../../Figures/03021102_bv_fig_11_1_log_barrier_approximation.jpeg" alt="Boyd and Vandenberghe Figure 11.1: logarithmic barrier approximation to an indicator" style="width: 60%; height: auto;">


**Numerical Example:**

The figure above shows barrier growth near a boundary; the arithmetic evaluates one interior point.

For the constraint $f_1(x)=x-2<0$, evaluate the point $x=1$. Then
$$
f_1(1)=1-2=-1.
$$

The point is strictly feasible because $-1<0$. The log-barrier contribution is
$$
-\log(-f_1(1))=-\log(-(-1))=-\log(1)=0.
$$

The finite value confirms that the point is in the strict interior.


```python
import sympy as sp

point = sp.Integer(1)
constraint_value = point - 2
barrier_value = -sp.log(-constraint_value)

barrier_value
```
