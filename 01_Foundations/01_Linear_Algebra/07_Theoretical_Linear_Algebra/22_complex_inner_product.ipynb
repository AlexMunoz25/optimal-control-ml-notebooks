### 1.1.7.22. Complex Inner Product


$$
\langle \mathbf{u}, \mathbf{v} \rangle = \mathbf{u}^{\mathsf{H}} \mathbf{v}
$$


**Explanation:**

A complex inner product modifies the dot product by conjugating one argument. This change is what makes orthogonality and norm behave correctly over $\mathbb{C}$, and it depends on the conjugation operation introduced in [Hermitian transpose](./22_hermitian_transpose.ipynb).

**Example:**

Let

$$
\mathbf{u}=\begin{bmatrix}1+i\\2-i\end{bmatrix},
\qquad
\mathbf{v}=\begin{bmatrix}2\\i\end{bmatrix}.
$$

Using conjugation on the left argument,

$$
\langle \mathbf{u},\mathbf{v}\rangle
=(1-i)2+(2+i)i.
$$

Compute each term:

$$
(1-i)2=2-2i,
\qquad
(2+i)i=2i+i^2=2i-1.
$$

Therefore

$$
\langle \mathbf{u},\mathbf{v}\rangle=(2-2i)+(2i-1)=1.
$$

Also

$$
\langle \mathbf{u},\mathbf{u}\rangle=|1+i|^2+|2-i|^2=2+5=7.
$$


```python
import sympy as sp

vector_u = sp.Matrix([1 + sp.I, 2 - sp.I])
vector_v = sp.Matrix([2, sp.I])
inner_product = sp.expand(vector_u.dot(vector_v, hermitian=True, conjugate_convention="left"))
self_inner_product = sp.expand(vector_u.dot(vector_u, hermitian=True, conjugate_convention="left"))

print("inner_product =", inner_product)
print("self_inner_product =", self_inner_product)
```
