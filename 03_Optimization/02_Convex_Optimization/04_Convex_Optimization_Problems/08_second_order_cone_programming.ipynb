### 3.2.4.8. Second-Order Cone Programming


$$\|A_ix+b_i\|_2\leq c_i^Tx+d_i$$


**Explanation:**

A second-order cone constraint bounds a norm by a scalar component. SOCPs collect affine equalities and inequalities of this cone form, which covers many robust, quadratic, and norm-constrained models.

The left side is a Euclidean norm of an affine expression in $x$. The right side is another affine expression, so the constraint says that a [norm](../../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb) must fit under a scalar affine bound.

**Properties:**
- Norm balls are convex by the triangle inequality.
- Norm cones contain pairs whose scalar component dominates the norm.
- Second-order cone constraints use this pattern.

**Intuition:**

The example uses a three-four-five triangle, so the norm is exact.


**Numerical Example:**

Let
$$
u=\begin{bmatrix}3\\4\end{bmatrix},\quad t=5.
$$

Compute the norm:
$$
\|u\|_2=\sqrt{3^2+4^2}=\sqrt{9+16}=\sqrt{25}=5.
$$

The cone test is
$$
\|u\|_2\leq t.
$$

Substitution gives
$$
5\leq5,
$$
so the point is feasible and lies on the cone boundary.


```python
import sympy as sp

vector = sp.Matrix([3, 4])
height = sp.Integer(5)
norm_value = sp.sqrt((vector.T * vector)[0])

norm_value, norm_value <= height
```
