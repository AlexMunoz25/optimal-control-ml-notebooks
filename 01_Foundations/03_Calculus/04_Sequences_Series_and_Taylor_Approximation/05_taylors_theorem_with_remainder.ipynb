### 1.3.4.5. Taylor's Theorem with Remainder

$$
f(x) = p_n(x) + R_n(x), \qquad
R_n(x) = \frac{f^{(n+1)}(c)}{(n+1)!}(x - a)^{n+1}
$$

$$
\text{for some } c \text{ between } a \text{ and } x; \qquad
|R_n(x)| \le \frac{M}{(n+1)!}|x - a|^{n+1}\ \ \big(|f^{(n+1)}| \le M\big).
$$

**Explanation:**

Taylor's theorem makes the [Taylor approximation](./04_taylor_and_maclaurin_series.ipynb) quantitative: the exact function equals the degree-$n$ Taylor polynomial plus a remainder $R_n$ given in Lagrange form by the $(n+1)$th derivative at some unknown intermediate point $c$. Bounding that derivative by $M$ yields a concrete error bound that shrinks like $|x-a|^{n+1}/(n+1)!$ — so the remainder is $O\big((x-a)^{n+1}\big)$ in [asymptotic notation](../01_Limits_and_Continuity/05_asymptotic_notation_big_o_little_o.ipynb). This is the result that lets second-order optimization methods trust a local quadratic model and that certifies linearization error in control.

**Properties:**
- The Lagrange remainder is proved from the [mean value theorem](../02_Single_Variable_Differentiation/07_mean_value_theorem.ipynb) (its $n=0$ case).
- A Taylor series converges to $f$ on $I$ exactly when $R_n(x) \to 0$ for every $x \in I$.

**Numerical Example:**

Approximate $e^{0.1}$ with the degree-2 Maclaurin polynomial $p_2(x) = 1 + x + \tfrac{x^2}{2}$, so $p_2(0.1) = 1.105$. The remainder is

$$
R_2(0.1) = \frac{e^{c}}{3!}(0.1)^3, \qquad 0 < c < 0.1 .
$$

Since $e^c \le e^{0.1} < 1.2 =: M$ on $[0, 0.1]$, the error is bounded by

$$
|R_2(0.1)| \le \frac{1.2}{6}(0.1)^3 = 2.0\times 10^{-4} .
$$

The true error is $e^{0.1} - 1.105 = 1.7\times 10^{-4}$, comfortably within the bound.

```python
import sympy as sp

x = sp.symbols("x")
f = sp.exp(x)
center, evaluation_point, degree = 0, sp.Rational(1, 10), 2

taylor_polynomial = sp.series(f, x, center, degree + 1).removeO()
approximate_value = taylor_polynomial.subs(x, evaluation_point)
true_value = f.subs(x, evaluation_point)
true_error = sp.N(true_value - approximate_value)

bound_M = sp.Rational(12, 10)
remainder_bound = bound_M / sp.factorial(degree + 1) * evaluation_point**(degree + 1)

print("p_2(x)        =", taylor_polynomial)
print("p_2(0.1)      =", float(approximate_value))
print("e^0.1 (true)  =", sp.N(true_value, 8))
print("true error    =", true_error)
print("Lagrange bound |R_2| <=", sp.N(remainder_bound))
```
