### 3.2.7.10. Optimal Experiment Design


$$\min_\lambda\Phi\left(\sum_i\lambda_iv_iv_i^T\right),\qquad\lambda_i\geq0,
\sum_i\lambda_i=1$$


**Explanation:**

Optimal experiment design chooses nonnegative experiment weights to shape an information matrix. Convex scalarizations of this matrix express different uncertainty criteria. Feasibility is the simplex of experiment weights, and optimality improves the chosen matrix measure.

The variables $\lambda_i$ are nonnegative experiment weights that sum to one. The matrix $\sum_i\lambda_i v_i v_i^T$ is a weighted sum of [outer products](../../../01_Foundations/01_Linear_Algebra/03_Matrix/14_matrix_outer_product.ipynb), and $\Phi$ scores that information matrix.

**Properties:**
- Information matrices are weighted sums of outer products.
- Experiment weights are nonnegative and sum to one.
- D-, A-, and E-optimality use different scalarizations.

**Intuition:**

The example builds a small information matrix and computes a [determinant](../../../01_Foundations/01_Linear_Algebra/03_Matrix/10_matrix_determinant.ipynb) criterion.


**Numerical Example:**

Let
$$
v_1=\begin{bmatrix}1\\0\end{bmatrix},\qquad
v_2=\begin{bmatrix}1\\1\end{bmatrix},
$$
with weights $1/2$ and $1/2$. Then
$$
M=\frac12v_1v_1^T+\frac12v_2v_2^T.
$$

The outer products are
$$
v_1v_1^T=\begin{bmatrix}1&0\\0&0\end{bmatrix},\qquad
v_2v_2^T=\begin{bmatrix}1&1\\1&1\end{bmatrix}.
$$

Therefore
$$
M=\begin{bmatrix}1&1/2\\1/2&1/2\end{bmatrix},
$$
and
$$
\det M=1\cdot\frac12-\frac12\cdot\frac12=\frac14.
$$


```python
import sympy as sp

vector_1 = sp.Matrix([1, 0])
vector_2 = sp.Matrix([1, 1])
information_matrix = sp.Rational(1, 2) * vector_1 * vector_1.T + sp.Rational(1, 2) * vector_2 * vector_2.T
determinant = information_matrix.det()

print("information_matrix =", information_matrix)
print("determinant =", determinant)
```
