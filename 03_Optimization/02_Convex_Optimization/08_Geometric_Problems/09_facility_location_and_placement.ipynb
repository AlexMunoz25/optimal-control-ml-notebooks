### 3.2.8.9. Facility Location and Placement


$$\min_x\sum_iw_i\|x-a_i\|_2$$


**Explanation:**

Facility location chooses a point that minimizes weighted distances to demand locations. [Norm](../../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb)-[distance](../../../01_Foundations/01_Linear_Algebra/05_Geometrical_Aspects_of_Linear_Algebra/04_distance_between_points.ipynb) objectives are convex in the facility location. Feasibility describes allowed placements, while optimality compares total distance or weighted distance.

The decision vector $x$ is the facility location, $a_i$ are demand locations, and $w_i$ are nonnegative weights. The objective sums weighted Euclidean distances from the facility to each demand point.

**Properties:**
- One-dimensional absolute-distance models are minimized by median-like locations.
- Weights scale individual demand contributions.
- Norm objectives preserve convexity.

**Intuition:**

The example compares two candidate locations by total distance.


**Numerical Example:**

Demand locations are $0$, $2$, and $6$. At candidate $x=2$,
$$
|2-0|+|2-2|+|2-6|=2+0+4=6.
$$

At candidate $x=3$,
$$
|3-0|+|3-2|+|3-6|=3+1+3=7.
$$

The smaller total distance is attained by $x=2$ among these two candidates.


```python
import sympy as sp

locations = [sp.Integer(0), sp.Integer(2), sp.Integer(6)]
candidate_1 = sp.Integer(2)
candidate_2 = sp.Integer(3)
distance_1 = sum(abs(candidate_1 - location) for location in locations)
distance_2 = sum(abs(candidate_2 - location) for location in locations)

distance_1, distance_2
```
