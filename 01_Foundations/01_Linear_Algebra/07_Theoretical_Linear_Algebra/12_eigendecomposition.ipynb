### 1.1.7.12. Eigendecomposition


$$
A = Q \Lambda Q^{-1}
$$


**Explanation:**

An eigendecomposition writes a diagonalizable matrix in a basis of eigenvectors. The diagonal matrix $\Lambda$ contains the [eigenvalues](./01_eigenvalue.ipynb), while the columns of $Q$ are [eigenvectors](./02_eigenvector.ipynb). The basis interpretation itself is developed in [change of basis to the eigenbasis](./05_change_of_basis_to_eigenbasis.ipynb).

**Example:**

For

$$
A =
\begin{bmatrix}
9 & -2 \\
-2 & 6
\end{bmatrix},
$$

one eigendecomposition is

$$
Q =
\begin{bmatrix}
1 & 2 \\
2 & -1
\end{bmatrix},
\qquad
\Lambda =
\begin{bmatrix}
5 & 0 \\
0 & 10
\end{bmatrix}.
$$


```python
import sympy as sp

matrix_a = sp.Matrix([[9, -2], [-2, 6]])
eigenvector_data = matrix_a.eigenvects()
matrix_q = sp.Matrix.hstack(eigenvector_data[0][2][0], eigenvector_data[1][2][0])
matrix_lambda = sp.diag(eigenvector_data[0][0], eigenvector_data[1][0])

print("matrix_q =")
sp.pprint(matrix_q)
print("matrix_lambda =")
sp.pprint(matrix_lambda)
print("reconstruction =")
sp.pprint(sp.simplify(matrix_q * matrix_lambda * matrix_q.inv()))
```
