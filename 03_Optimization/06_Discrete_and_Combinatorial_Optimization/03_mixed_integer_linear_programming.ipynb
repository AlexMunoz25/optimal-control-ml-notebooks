### 3.6.3. Mixed-Integer Linear Programming (MILP)

$$
\begin{aligned}
\min_{\mathbf{x},\mathbf{y}}\quad & \mathbf{c}^\top \mathbf{x} + \mathbf{d}^\top \mathbf{y} \\
\text{subject to}\quad & A\mathbf{x} + B\mathbf{y} \le \mathbf{b},\\
& \mathbf{x}\in\mathbb{R}^{n}_{\ge 0},\quad \mathbf{y}\in\mathbb{Z}^{q}_{\ge 0}.
\end{aligned}
$$

**Explanation:**

A mixed-integer linear program (MILP) has a linear objective and constraints but only *some* variables are required to be integer, the rest remaining continuous. This mixture models decisions where discrete choices (how many machines, whether to open a facility) interact with continuous quantities (how many hours, how much material). Because the continuous part is an LP once the integers are fixed, MILP is solved by branching on the integer variables and solving an [LP relaxation](../04_Convex_Problem_Classes/01_linear_programming.ipynb) at each node.

**Numerical Example:**

A task needs $10$ units of output from continuous hours $x\in[0,8]$ and an integer number of machines $y\in\{0,1,2,3\}$, each machine supplying $4$ units:
$$
\min\; 2x + 5y
\quad\text{s.t.}\quad x + 4y \ge 10,\; 0\le x\le 8,\; y\in\{0,1,2,3\}.
$$

For each integer $y$ the cheapest continuous $x$ is $\max(0,\,10-4y)$ (feasible only if this is $\le 8$):
$$
y=1:\; x=6,\ \text{cost }17;\quad
y=2:\; x=2,\ \text{cost }14;\quad
y=3:\; x=0,\ \text{cost }15.
$$
The optimum is $y=2,\ x=2$ with cost $14$; $y=0$ is infeasible because it would need $x=10>8$.

```python
machine_costs = {}
for machines in range(0, 4):
    required_hours = max(0, 10 - 4*machines)
    if required_hours <= 8:
        machine_costs[machines] = 2*required_hours + 5*machines

best_machines = min(machine_costs, key=machine_costs.get)
best_hours = max(0, 10 - 4*best_machines)

print("cost per machine count:", machine_costs)
print("optimum: hours =", best_hours, " machines =", best_machines, " cost =", machine_costs[best_machines])
```

**Equivalent scipy.milp implementation:**

```python
import numpy as np
from scipy.optimize import milp, LinearConstraint, Bounds

cost = np.array([2.0, 5.0])
demand = LinearConstraint(np.array([[1.0, 4.0]]), 10, np.inf)
result = milp(c=cost, constraints=demand, integrality=np.array([0, 1]),
              bounds=Bounds(lb=[0, 0], ub=[8, 3]))

print("optimum: hours =", result.x[0], " machines =", result.x[1], " cost =", result.fun)
```
