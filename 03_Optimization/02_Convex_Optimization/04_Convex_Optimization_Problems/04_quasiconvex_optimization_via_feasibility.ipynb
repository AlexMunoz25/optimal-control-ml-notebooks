### 3.2.4.4. Quasiconvex Optimization via Feasibility


$$f_0(x)\leq t$$


**Explanation:**

Quasiconvex optimization is usually handled through a sequence of convex feasibility tests. For a proposed objective level, the sublevel constraint is convex; bisection then searches for the smallest feasible level.

The inequality fixes a candidate objective level $t$. Solving the associated feasibility problem asks whether there exists an $x$ whose objective value is no larger than that level.

**Properties:**
- Convex functions have convex epigraphs.
- Quasiconvex functions have convex sublevel sets.
- Epigraph transformations replace objectives by constraints.

**Intuition:**

The example checks the same point against two different levels.


**Numerical Example:**

Let $f(x)=x^2$ and $x=2$. First compute
$$
f(2)=2^2=4.
$$

At level $5$,
$$
4\leq5,
$$
so $x=2$ is in the sublevel set. At level $3$,
$$
4\leq3
$$
is false, so $x=2$ is not in that lower sublevel set.


```python
import sympy as sp

function_value = sp.Integer(2) ** 2

function_value <= 5, function_value <= 3
```
