### 3.1.8. Gradient Method Framework


$$
x^{k+1}=x^k-\alpha^k D^k \nabla f(x^k),
\qquad
D^k \succ 0.
$$


**Explanation:**

Bertsekas uses the term gradient method broadly: the search direction is related to the gradient and is chosen so that it is a descent direction whenever the current point is not stationary. The matrix $D^k$ defines the local scaling of the gradient step. With $D^k=I$, the method is steepest descent; with diagonal $D^k$, it becomes a scaled steepest descent method; with $D^k=(\nabla^2 f(x^k))^{-1}$ and a positive definite Hessian, it becomes Newton's direction. The descent property follows from [positive definiteness](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/17_positive_definite_matrix.ipynb), since $\nabla f(x^k)^{\mathsf T}D^k\nabla f(x^k)>0$ for a nonzero gradient.

This framework is important because it lets several algorithms be compared using the same notation. Instead of memorizing methods separately, the reader can ask what information is used to build $D^k$ and how expensive that information is to obtain. The later rate results then explain why better scaling can dramatically improve performance.

**Example:**

Steepest descent can zig-zag on elongated level sets because the negative gradient points across the valley rather than along it. Newton's method uses a quadratic model and curvature information to point more directly toward the minimizer near a nonsingular solution. The code cell contrasts an ordinary gradient step with a curvature-scaled step on an ill-conditioned quadratic.

<img src="../../Figures/030101_berts_fig_1_2_4_slow_steepest_descent.jpeg" alt="Bertsekas Figure 1.2.4: slow steepest descent on elongated level sets" style="width: 60%; height: auto;">

<img src="../../Figures/030101_berts_fig_1_2_5_newton_quadratic_model.jpeg" alt="Bertsekas Figure 1.2.5: Newton step from a quadratic approximation" style="width: 60%; height: auto;">


```python
import numpy as np

Q = np.array([[20.0, 0.0], [0.0, 1.0]])
initial_point = np.array([1.0, 1.0])
stepsize = 0.08
steepest_point = initial_point.copy()
scaled_point = initial_point.copy()
scaling_matrix = np.linalg.inv(Q)

for iteration in range(6):
    steepest_point = steepest_point - stepsize * (Q @ steepest_point)
    scaled_point = scaled_point - (scaling_matrix @ (Q @ scaled_point))

steepest_point, scaled_point
```
