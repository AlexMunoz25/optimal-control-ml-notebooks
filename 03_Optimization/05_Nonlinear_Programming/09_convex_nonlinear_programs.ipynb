### 3.5.9. Convex Nonlinear Programs

$$
\begin{aligned}
\min_{\mathbf{x}\in\mathbb{R}^n}\quad & f(\mathbf{x}) \\
\text{subject to}\quad & g_i(\mathbf{x}) \le 0,\quad i=1,\dots,m,\\
& A\mathbf{x} = \mathbf{b},
\end{aligned}
$$
with $f$ and every $g_i$ convex.

**Explanation:**

A convex nonlinear program keeps nonlinear objective and constraints but requires $f$ and the $g_i$ to be convex and the equality constraints to be affine. This structure guarantees that the feasible set is convex and that **every local minimizer is a global minimizer**, so the KKT conditions become *sufficient*, not merely necessary. This is the property that makes the problem classes of [convex optimization](../03_Convex_Optimization/29_convex_problem_standard_form_and_optimality.ipynb) reliably solvable; nonconvex programs, treated [next](./10_nonconvex_nonlinear_programs.ipynb), lose this guarantee.

**Numerical Example:**

Minimize an exponential-plus-quadratic cost on a halfspace:
$$
\min_{x_1,x_2}\; e^{x_1} + x_1^2 + x_2^2
\quad\text{subject to}\quad x_1 + x_2 \ge 1.
$$

Both the objective (sum of convex functions) and the constraint (affine) are convex. At the solution the constraint is active, $x_1 + x_2 = 1$, and eliminating $x_2 = 1 - x_1$ gives the one-variable convex problem
$$
\min_{x_1}\; e^{x_1} + x_1^2 + (1-x_1)^2,
$$
whose derivative $e^{x_1} + 4x_1 - 2 = 0$ has the unique root $x_1 \approx 0.234$, so $\mathbf{x}^\star \approx (0.234, 0.766)$ is the global minimizer.

```python
import sympy as sp

x_1 = sp.symbols("x_1", real=True)
reduced_objective = sp.exp(x_1) + x_1**2 + (1 - x_1)**2
stationarity = sp.diff(reduced_objective, x_1)

minimizer_x1 = sp.nsolve(stationarity, x_1, 0.0)
minimizer = (float(minimizer_x1), float(1 - minimizer_x1))
minimum_value = float(reduced_objective.subs(x_1, minimizer_x1))

print("stationarity condition:", stationarity, "= 0")
print("minimizer =", minimizer)
print("minimum value =", minimum_value)
```

**Equivalent casadi implementation:**

```python
import casadi as ca

decision_variable = ca.SX.sym("x", 2)
objective = ca.exp(decision_variable[0]) + decision_variable[0]**2 + decision_variable[1]**2
constraint = decision_variable[0] + decision_variable[1]

solver = ca.nlpsol("solver", "ipopt", {"x": decision_variable, "f": objective, "g": constraint},
                   {"ipopt.print_level": 0, "print_time": 0, "ipopt.sb": "yes"})
solution = solver(x0=[0, 0], lbg=1, ubg=ca.inf)

print("minimizer =", solution["x"])
print("minimum value =", float(solution["f"]))
```
