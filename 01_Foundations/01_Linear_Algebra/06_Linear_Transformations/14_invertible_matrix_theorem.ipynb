### 1.1.6.14. Invertible Matrix Theorem


$$
A \text{ invertible }
\Longleftrightarrow
\mathcal{N}(A) = \{ \mathbf{0} \}
\Longleftrightarrow
\operatorname{rank}(A) = n
\Longleftrightarrow
\det(A) \neq 0
$$


**Explanation:**

The invertible matrix theorem collects several equivalent ways to recognize when a square matrix has an inverse. A trivial [kernel](./03_kernel.ipynb), a full [image space](./02_image_space.ipynb), and an [invertible linear transformation](./07_invertible_linear_transformation.ipynb) are all different descriptions of the same situation.

**Example:**

For

$$
A =
\begin{bmatrix}
2 & 1 \\
1 & 1
\end{bmatrix},
$$

the determinant is $1$, so the matrix is invertible and all the equivalent statements hold.


```python
import sympy as sp

matrix_a = sp.Matrix([[2, 1], [1, 1]])
determinant_value = matrix_a.det()
rank_value = matrix_a.rank()
reduced_matrix, _ = matrix_a.rref()

print("determinant_value =", determinant_value)
print("rank_value =", rank_value)
print("reduced_matrix =")
sp.pprint(reduced_matrix)
```
