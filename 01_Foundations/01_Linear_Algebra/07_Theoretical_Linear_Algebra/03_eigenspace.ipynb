### 1.1.7.3. Eigenspace


$$
E_{\lambda} = \mathcal{N}(A - \lambda I)
$$


**Explanation:**

The eigenspace of $\lambda$ is the null space of $A - \lambda I$. It is not the same as a single [eigenvector](./02_eigenvector.ipynb): an eigenspace can contain many eigenvectors and has its own dimension, the geometric multiplicity. This viewpoint becomes especially useful when interpreting [normal matrices](./08_normal_matrix.ipynb).

**Example:**

If $\lambda$ is repeated, the eigenspace may contain several linearly independent eigenvectors. Its dimension is the geometric multiplicity of $\lambda$.


```python
import sympy as sp

matrix_a = sp.Matrix([[9, -2], [-2, 6]])
eigenspace_basis = (matrix_a - 5 * sp.eye(2)).nullspace()

print("eigenspace_basis =", eigenspace_basis)
```
