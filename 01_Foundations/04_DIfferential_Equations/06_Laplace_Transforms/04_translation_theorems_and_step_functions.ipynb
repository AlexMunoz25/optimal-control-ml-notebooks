### 1.4.6.4. Translation Theorems and Step Functions

$$
\textbf{First (s-shift):}\ \ \mathcal{L}\{e^{at}f(t)\} = F(s - a);
\qquad
\textbf{Second (t-shift):}\ \ \mathcal{L}\{f(t-a)\,\mathcal{U}(t-a)\} = e^{-as}F(s) .
$$

**Explanation:**

Two translation (shifting) theorems extend the transform table. The **first** says multiplying $f(t)$ by $e^{at}$ shifts its transform on the $s$-axis, $F(s) \to F(s-a)$. The **unit step function** $\mathcal{U}(t-a)$ — zero before $a$, one after — "turns on" a function at $t = a$ and writes piecewise forcing compactly. The **second** translation theorem gives the transform of a time-shifted, switched-on function, $\mathcal{L}\{f(t-a)\mathcal{U}(t-a)\} = e^{-as}F(s)$, with the special case $\mathcal{L}\{\mathcal{U}(t-a)\} = e^{-as}/s$. These handle the switched and delayed inputs (step commands, piecewise loads) that pervade control and signal models.

**Intuition:**

The unit step $\mathcal{U}(t-a)$ jumps from $0$ to $1$ at $t = a$, switching a term on.

**Numerical Example:**

**First translation.** With $\mathcal{L}\{t^3\} = 6/s^4$, shifting by $a = 5$:

$$
\mathcal{L}\{e^{5t}t^3\} = \frac{6}{s^4}\bigg|_{s \to s-5} = \frac{6}{(s-5)^4}.
$$

With $\mathcal{L}\{\cos 4t\} = s/(s^2+16)$, shifting by $a = -2$:

$$
\mathcal{L}\{e^{-2t}\cos 4t\} = \frac{s+2}{(s+2)^2 + 16}.
$$

**Second translation / step.** The transform of the unit step is

$$
\mathcal{L}\{\mathcal{U}(t-a)\} = \int_a^{\infty} e^{-st}\,dt = \frac{e^{-as}}{s},
$$

and a function switched on at $t = a$ transforms as $\mathcal{L}\{f(t-a)\mathcal{U}(t-a)\} = e^{-as}F(s)$ — the factor $e^{-as}$ encodes the time delay.

```python
import sympy as sp

t, s = sp.symbols("t s", positive=True)
a = sp.symbols("a", positive=True)

first_a = sp.laplace_transform(sp.exp(5 * t) * t**3, t, s, noconds=True)
first_b = sp.laplace_transform(sp.exp(-2 * t) * sp.cos(4 * t), t, s, noconds=True)
unit_step = sp.laplace_transform(sp.Heaviside(t - a), t, s, noconds=True)

print("L{e^5t t^3}      =", first_a)
print("L{e^-2t cos 4t}  =", sp.simplify(first_b))
print("L{U(t-a)}        =", unit_step)
```
