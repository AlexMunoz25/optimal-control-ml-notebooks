### A2.2.2. LLVM Intermediate Representation Instructions

> *LLVM instructions are the operations that transform SSA values. Each instruction produces at most one result value and belongs to exactly one basic block.*

**Explanation:**

LLVM IR instructions fall into several categories:

**Arithmetic:**
- `add`, `sub`, `mul`, `sdiv`, `udiv` — integer arithmetic.
- `fadd`, `fsub`, `fmul`, `fdiv` — floating-point arithmetic.

**Memory:**
- `alloca` — allocate stack memory, returns a pointer.
- `load` — read a value from a pointer.
- `store` — write a value to a pointer.
- `getelementptr` (GEP) — compute an address offset without accessing memory.

**Control Flow (Terminators):**
- `br` — unconditional or conditional branch.
- `ret` — return from function.
- `switch` — multi-way branch.
- `unreachable` — marks a point that should never be reached.

**Conversion:**
- `sext`, `zext`, `trunc` — integer sign-extend, zero-extend, truncate.
- `sitofp`, `fptosi` — integer ↔ float conversions.
- `bitcast`, `inttoptr`, `ptrtoint` — reinterpret bits.

**Other:**
- `icmp`, `fcmp` — integer/float comparison, returns `i1`.
- `phi` — SSA phi function (selects value based on predecessor block).
- `call` — invoke a function.
- `select` — ternary operator (`cond ? a : b`).

**Example:**

```llvm
define i32 @max(i32 %a, i32 %b) {
  %cmp = icmp sgt i32 %a, %b
  %result = select i1 %cmp, i32 %a, i32 %b
  ret i32 %result
}
```

````python
instruction_categories = {
    "Arithmetic": ["add", "sub", "mul", "sdiv", "fadd", "fsub", "fmul", "fdiv"],
    "Memory": ["alloca", "load", "store", "getelementptr"],
    "Terminators": ["br", "ret", "switch", "unreachable"],
    "Conversion": ["sext", "zext", "trunc", "sitofp", "fptosi", "bitcast"],
    "Comparison": ["icmp", "fcmp"],
    "Other": ["phi", "call", "select"],
}

max_function_ir = """
define i32 @max(i32 %a, i32 %b) {
  %cmp = icmp sgt i32 %a, %b
  %result = select i1 %cmp, i32 %a, i32 %b
  ret i32 %result
}
"""

print("LLVM Instruction Categories:")
for category, instructions in instruction_categories.items():
    print(f"  {category}: {', '.join(instructions)}")

print("\nExample — max function:")
print(max_function_ir)
````
