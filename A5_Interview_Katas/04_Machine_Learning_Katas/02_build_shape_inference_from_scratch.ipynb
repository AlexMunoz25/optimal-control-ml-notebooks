### A5.4.2. Build Shape Inference from Scratch

**Problem:**

Build a system that, given the shapes of input tensors and an operation name, computes the output tensor shape without executing the operation.

**Explanation:**

Shape inference determines output shapes from input shapes and operation semantics. This lets you validate a computation graph and allocate memory before running any computation.

How to build it:

1. Define a dictionary mapping operation names to shape-inference functions.
2. Each shape-inference function takes a list of input shapes and returns the output shape.
3. For element-wise operations (add, multiply): the output shape equals the input shape (inputs must match).
4. For matrix multiplication: inputs are `(M, K)` and `(K, N)`, output is `(M, N)`.
5. For reduce operations: one dimension is removed from the shape.
6. Look up the operation name in the dictionary and call the corresponding function.

**Example:**

`infer_shape("matmul", [(3, 4), (4, 5)])` returns `(3, 5)`.

```python
def infer_elementwise(input_shapes):
    shape_a, shape_b = input_shapes
    return shape_a


def infer_matmul(input_shapes):
    shape_a, shape_b = input_shapes
    return (shape_a[0], shape_b[1])


def infer_reduce(input_shapes, axis):
    shape = list(input_shapes[0])
    del shape[axis]
    return tuple(shape)


def infer_reshape(input_shapes, new_shape):
    return tuple(new_shape)


SHAPE_RULES = {
    "add": infer_elementwise,
    "multiply": infer_elementwise,
    "matmul": infer_matmul,
}


def infer_shape(operation, input_shapes, **kwargs):
    if operation == "reduce_sum":
        return infer_reduce(input_shapes, kwargs["axis"])
    if operation == "reshape":
        return infer_reshape(input_shapes, kwargs["new_shape"])
    return SHAPE_RULES[operation](input_shapes)


print(f"add: (3,4) + (3,4) → {infer_shape('add', [(3, 4), (3, 4)])}")
print(f"matmul: (3,4) @ (4,5) → {infer_shape('matmul', [(3, 4), (4, 5)])}")
print(f"reduce_sum axis=1: (3,4) → {infer_shape('reduce_sum', [(3, 4)], axis=1)}")
print(f"reshape: (2,6) → {infer_shape('reshape', [(2, 6)], new_shape=(3, 4))}")
```
