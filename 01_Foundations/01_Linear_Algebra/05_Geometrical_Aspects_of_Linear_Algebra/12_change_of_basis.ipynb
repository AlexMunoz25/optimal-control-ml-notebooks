### 1.1.5.12. Change of Basis


$$
\left[ \mathbf{v} \right]_{B'} = {}_{B'}\!\left[ \mathbb{1} \right]_B \left[ \mathbf{v} \right]_B
$$


**Explanation:**

A change of basis leaves the geometric vector fixed and changes only the coefficient column used to describe it. The change-of-basis matrix can be viewed as the identity transformation written between two coordinate frames, and that same idea is central to [similarity transformations](../06_Linear_Transformations/13_similarity_transformation.ipynb).

</p>

**Example:**

If the new basis is a $45^\circ$ rotation of the standard basis, then

$$
{}_{B'}\!\left[ \mathbb{1} \right]_B =
\begin{bmatrix}
\cos \theta & \sin \theta \\
-\sin \theta & \cos \theta
\end{bmatrix},
\qquad
\theta = \frac{\pi}{4}.
$$


```python
import sympy as sp

rotation_angle = sp.pi / 4
change_of_basis_matrix = sp.Matrix(
    [
        [sp.cos(rotation_angle), sp.sin(rotation_angle)],
        [-sp.sin(rotation_angle), sp.cos(rotation_angle)],
    ]
)
vector_in_old_basis = sp.Matrix([3, 1])
vector_in_new_basis = sp.simplify(change_of_basis_matrix * vector_in_old_basis)

print("change_of_basis_matrix =")
sp.pprint(change_of_basis_matrix)
print("vector_in_new_basis =", vector_in_new_basis)
```
