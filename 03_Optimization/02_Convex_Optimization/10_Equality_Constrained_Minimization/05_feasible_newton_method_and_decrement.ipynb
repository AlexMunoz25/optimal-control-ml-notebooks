### 3.2.10.5. Feasible Newton Method and Decrement


$$\lambda(x)^2=\Delta x^T\nabla^2f(x)\Delta x$$


**Explanation:**

The feasible Newton method keeps iterates on the equality constraint and uses the constrained Newton decrement to measure predicted decrease. Feasibility is the equality constraint or its linearized version; optimality is stationarity together with that feasibility condition.

**Properties:**
- Equality constraints introduce unrestricted multipliers.
- KKT systems couple stationarity and feasibility.
- Eliminating constraints can reduce the number of variables.

**Intuition:**

The numerical example keeps the equality constraint visible in every algebraic step.


**Numerical Example:**

Using the same constrained Newton step,
$$
\Delta x=\begin{bmatrix}1/3\\-1/3\end{bmatrix},
\quad
H=\begin{bmatrix}2&0\\0&4\end{bmatrix},
$$
the squared decrement is
$$
\lambda^2=\Delta x^TH\Delta x.
$$

Compute
$$
H\Delta x=\begin{bmatrix}2/3\\-4/3\end{bmatrix},
$$
so
$$
\lambda^2=(1/3)(2/3)+(-1/3)(-4/3)=2/9+4/9=2/3.
$$


```python
import sympy as sp

step = sp.Matrix([sp.Rational(1, 3), sp.Rational(-1, 3)])
hessian = sp.diag(2, 4)
decrement_squared = (step.T * hessian * step)[0]

decrement_squared
```
