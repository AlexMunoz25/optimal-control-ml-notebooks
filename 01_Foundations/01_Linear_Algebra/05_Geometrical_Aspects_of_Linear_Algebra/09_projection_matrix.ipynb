### 1.1.5.9. Projection Matrix


$$
\mathbf{v} = M_{\Pi} \mathbf{u}
$$

$$
M_{\Pi}^2 = M_{\Pi}
$$


**Explanation:**

Once a projection is recognized as a linear transformation, it can be implemented by a matrix. Savov highlights idempotence as the defining algebraic property: applying the same projection twice does nothing new. This notebook depends on the geometric notebooks on [projection onto a line](./07_projection_onto_a_line.ipynb) and [projection onto a plane](./08_projection_onto_a_plane.ipynb).

**Properties:**
- The eigenvalues of a projection matrix are $0$ and $1$.
- Orthogonal projection onto a unit vector $\hat{\mathbf{a}}$ uses $\hat{\mathbf{a}} \hat{\mathbf{a}}^{\mathsf{T}}$.

**Example:**

For the projection onto the $xy$-plane,

$$
M_{\Pi} =
\begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 0
\end{bmatrix},
$$

so

$$
M_{\Pi}
\begin{bmatrix}
1 \\
2 \\
3
\end{bmatrix}
=
\begin{bmatrix}
1 \\
2 \\
0
\end{bmatrix}.
$$


```python
import sympy as sp

projection_matrix = sp.Matrix([[1, 0, 0], [0, 1, 0], [0, 0, 0]])
vector_u = sp.Matrix([1, 2, 3])
single_projection = projection_matrix * vector_u
double_projection = projection_matrix * single_projection

print("single_projection =", single_projection)
print("double_projection =", double_projection)
print("idempotent =", double_projection == single_projection)
```
