### A2.1.2. Control Flow Graph

> *A control flow graph (CFG) is a directed graph where each node represents a basic block and each edge represents a possible transfer of control between blocks.*

**Explanation:**

A **Control Flow Graph (CFG)** is the fundamental data structure for compiler analysis and optimization. It decomposes a function into **basic blocks** â€” maximal sequences of instructions with:

- A single entry point (only the first instruction can be reached from outside the block).
- A single exit point (only the last instruction can transfer control to another block).

Edges in the CFG represent branches: conditional branches produce two successor edges, unconditional branches produce one, and return instructions produce none.

**Properties:**

- **Entry block** â€” the block where function execution begins.
- **Predecessors / Successors** â€” the blocks that can transfer control to/from a given block.
- **Dominance** â€” block A *dominates* block B if every path from entry to B passes through A.
- **Back edges** â€” edges pointing to a dominating block, indicating loops.

In LLVM, each `Function` contains a list of `BasicBlock` objects. The CFG is implicit: each block's terminator instruction (`br`, `ret`, `switch`) defines the successor edges.

**Example:**

```
entry:
  br i1 %cond, label %then, label %else

then:
  ...
  br label %merge

else:
  ...
  br label %merge

merge:
  ret void
```

This produces a diamond-shaped CFG: `entry` â†’ `{then, else}` â†’ `merge`.

````python
cfg = {
    "entry": ["then", "else"],
    "then": ["merge"],
    "else": ["merge"],
    "merge": [],
}

print("Control Flow Graph (adjacency list):")
for block, successors in cfg.items():
    successors_str = ", ".join(successors) if successors else "(exit)"
    print(f"  {block} -> {successors_str}")

predecessors = {block: [] for block in cfg}
for block, successors in cfg.items():
    for successor in successors:
        predecessors[successor].append(block)

print("\nPredecessors:")
for block, preds in predecessors.items():
    preds_str = ", ".join(preds) if preds else "(entry)"
    print(f"  {block} <- {preds_str}")
````
