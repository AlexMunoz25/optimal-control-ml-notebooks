### 3.2.1.4. Convex Optimization in Nonconvex Workflows


$$d^\star\leq p^\star$$


**Explanation:**

Convex problems often appear inside nonconvex workflows as relaxations, approximations, or bounding problems. A dual or relaxed value may not be attainable by the original model, but it gives a benchmark for how good any feasible nonconvex solution can be.

**Properties:**
- Weak duality gives a lower bound for minimization.
- A primal feasible point gives an upper bound.
- A zero gap proves optimality when both sides are feasible.

**Intuition:**

The example treats the numbers as certificates rather than as iterates.


**Numerical Example:**

Suppose a primal feasible point has value $p=4$ and a dual feasible point has value $d=3$. Weak duality gives
$$
d\leq p^\star\leq p.
$$

Substitute the values:
$$
3\leq p^\star\leq4.
$$

The certificate gap is
$$
p-d=4-3=1.
$$

The dual point proves that no feasible primal solution can have value below $3$, while the primal point proves the optimum is at most $4$.


```python
import sympy as sp

primal_value = sp.Integer(4)
bound_value = sp.Integer(3)
gap = primal_value - bound_value

bound_value <= primal_value, gap
```
