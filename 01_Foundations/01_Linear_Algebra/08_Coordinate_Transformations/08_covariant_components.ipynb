### 1.1.8.8. Covariant Components


$$
A_i = \mathbf{A} \cdot \mathbf{e}_i
$$


**Explanation:**

Covariant components are the perpendicular projections onto the basis directions in a non-orthogonal coordinate system. They do not multiply the original basis vectors directly to reconstruct the vector, which is why the [dual basis](./06_dual_basis.ipynb) becomes necessary. The defining idea is this perpendicular-projection viewpoint.

</p>

**Example:**

In an oblique coordinate system, covariant components behave differently from [contravariant components](./07_contravariant_components.ipynb) because perpendicular shadows and coefficient weights are no longer the same object.


```python
import sympy as sp

basis_e1 = sp.Matrix([1, 0])
basis_e2 = sp.Matrix([sp.Rational(1, 2), sp.sqrt(3) / 2])
vector_a = sp.Matrix([3, 2])
covariant_components = [sp.simplify(vector_a.dot(basis_vector)) for basis_vector in [basis_e1, basis_e2]]

print("covariant_components =", covariant_components)
```
