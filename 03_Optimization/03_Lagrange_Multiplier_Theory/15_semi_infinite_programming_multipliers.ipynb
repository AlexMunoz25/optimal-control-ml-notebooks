### 3.3.15. Semi-Infinite Programming Multipliers

$$
\min_x f(x) \quad \text{subject to} \quad g_j(x)\le 0,\ j\in J
$$

$$
\nabla f(x^*)+\sum_{j\in J^*}\mu_j^*\nabla g_j(x^*)=0,
\qquad |J^*|\le n
$$

**Explanation:**

A semi-infinite program has finitely many decision variables but infinitely many inequality constraints. Under quasiregularity and closedness of the cone generated by active constraint gradients, only finitely many active constraints are needed in the multiplier equation. In $\mathbb{R}^n$, the active multiplier set can be chosen with at most $n$ linearly independent gradients when the objective gradient is nonzero.

**Intuition:**

<img src="../../Figures/03030315_berts_fig_3_3_8_semi_infinite_constraints.jpeg" alt="Bertsekas Figure 3.3.8: semi-infinite linear constraints tangent to the unit circle" style="width: 60%; height: auto;">

The figure shows linear constraints indexed by an angle. At the optimum for the full circle case, a single active tangent line supplies the normal direction needed to balance the objective gradient.

**Numerical Example:**

Consider the full-circle case

$$
\min -(x_1+x_2)
\quad \text{subject to} \quad
x_1\cos\phi+x_2\sin\phi\le 1,
\quad \phi\in[0,2\pi].
$$

The optimum is the point on the unit disk farthest in the $(1,1)$ direction:

$$
x^*=\begin{bmatrix}1/\sqrt{2}\\1/\sqrt{2}\end{bmatrix}.
$$

The active constraint angle is

$$
\phi^*=\frac{\pi}{4},
$$

because

$$
x_1^*\cos(\pi/4)+x_2^*\sin(\pi/4)
=\frac{1}{\sqrt{2}}\frac{1}{\sqrt{2}}
+\frac{1}{\sqrt{2}}\frac{1}{\sqrt{2}}
=\frac{1}{2}+\frac{1}{2}=1.
$$

The objective gradient is

$$
\nabla f(x^*)=\begin{bmatrix}-1\\-1\end{bmatrix}.
$$

The active constraint gradient is

$$
\nabla g_{\phi^*}(x^*)=\begin{bmatrix}\cos(\pi/4)\\\sin(\pi/4)\end{bmatrix}
=\begin{bmatrix}1/\sqrt{2}\\1/\sqrt{2}\end{bmatrix}.
$$

The multiplier equation is

$$
\begin{bmatrix}-1\\-1\end{bmatrix}
+\mu
\begin{bmatrix}1/\sqrt{2}\\1/\sqrt{2}\end{bmatrix}=0.
$$

Thus

$$
-1+\frac{\mu}{\sqrt{2}}=0
\quad \Rightarrow \quad
\mu^*=\sqrt{2}.
$$

Only one active constraint is needed, consistent with the finite-support multiplier result.

```python
import sympy as sp

mu = sp.symbols('mu')

active_gradient = sp.Matrix([sp.sqrt(2) / 2, sp.sqrt(2) / 2])
gradient_cost = sp.Matrix([-1, -1])
stationarity = gradient_cost + mu * active_gradient
mu_solution = sp.solve(stationarity, [mu], dict=True)[0]
active_constraint_value = sp.sqrt(2) / 2 * sp.sqrt(2) / 2 + sp.sqrt(2) / 2 * sp.sqrt(2) / 2

result = {
    'active_constraint_value': sp.simplify(active_constraint_value),
    'mu_solution': mu_solution,
    'stationarity_at_solution': sp.simplify(stationarity.subs(mu_solution)),
}

for label, value in result.items():
    print(f"{label} =", value)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

active_gradient = ca.DM([ca.sqrt(2) / 2, ca.sqrt(2) / 2])
gradient_cost = ca.DM([-1, -1])
multiplier = float(-gradient_cost[0] / active_gradient[0])
stationarity = gradient_cost + multiplier * active_gradient

print("multiplier =", multiplier)
print("stationarity_at_solution =", stationarity)
```
