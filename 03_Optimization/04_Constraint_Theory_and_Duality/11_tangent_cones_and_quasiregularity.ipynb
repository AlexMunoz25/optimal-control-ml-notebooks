### 3.4.11. Tangent Cones and Quasiregularity

$$
T(x)=\left\{y\mid \exists x^k\in X,
\ x^k\to x,
\ \frac{x^k-x}{\|x^k-x\|}\to\frac{y}{\|y\|}\right\}\cup\{0\}
$$

$$
T(x^*)=V(x^*)
$$

**Explanation:**

The tangent cone describes the actual local [vector](../../01_Foundations/01_Linear_Algebra/02_Vector/01_vector_definition.ipynb) directions from which feasible points can approach $x$. The first-order feasible variation cone $V(x)$ is computed from gradients, but it can overestimate true feasible directions when constraints are irregular. A point is quasiregular when the actual tangent cone and the first-order cone agree, and this agreement is enough to recover Lagrange multipliers.

**Numerical Example:**

Let

$$
X=\{(x_1,x_2)\mid x_2\ge x_1^2\}
$$

and examine the boundary point

$$
x^*=\begin{bmatrix}0\\0\end{bmatrix}.
$$

Write the constraint as

$$
g(x)=x_1^2-x_2\le 0.
$$

The gradient at $x^*$ is

$$
\nabla g(x^*)=\begin{bmatrix}2x_1^*\\-1\end{bmatrix}
=\begin{bmatrix}0\\-1\end{bmatrix}.
$$

The first-order feasible variation cone is

$$
V(x^*)=\{y\mid \nabla g(x^*)^T y\le 0\}
=\{y\mid -y_2\le 0\}
=\{y\mid y_2\ge 0\}.
$$

Now take a direction

$$
y=\begin{bmatrix}1\\0\end{bmatrix}.
$$

It belongs to $V(x^*)$ because $y_2=0$. To see it is also tangent, use the feasible sequence

$$
x^k=\begin{bmatrix}t_k\\t_k^2\end{bmatrix},
\qquad t_k\downarrow 0.
$$

Each point is feasible because

$$
t_k^2=(t_k)^2.
$$

The normalized direction is

$$
\frac{x^k-x^*}{\|x^k-x^*\|}
=\frac{(t_k,t_k^2)}{\sqrt{t_k^2+t_k^4}}
=\frac{(1,t_k)}{\sqrt{1+t_k^2}}
\to
\begin{bmatrix}1\\0\end{bmatrix}.
$$

Thus this boundary direction is a true tangent, not just a formal first-order variation.

```python
import sympy as sp

t = sp.symbols('t', positive=True)

sequence_point = sp.Matrix([t, t**2])
normalized_direction = sp.simplify(sequence_point / sp.sqrt(sequence_point.dot(sequence_point)))
limit_direction = sp.Matrix([
    sp.limit(normalized_direction[0], t, 0, dir='+'),
    sp.limit(normalized_direction[1], t, 0, dir='+'),
])
gradient_constraint_at_boundary = sp.Matrix([0, -1])
first_order_product = gradient_constraint_at_boundary.dot(sp.Matrix([1, 0]))

result = {
    'normalized_direction': normalized_direction,
    'limit_direction': limit_direction,
    'first_order_product': first_order_product,
}

for label, value in result.items():
    print(f"{label} =", value)
```
