### 1.1.6.13. Similarity Transformation


$$
M = P N P^{-1}
$$


**Explanation:**

A similarity transformation is the square-matrix version of [change of basis for matrix representations](./12_change_of_basis_for_matrix_representations.ipynb). The matrix $P$ translates between coordinate frames, so similar matrices represent the same [linear transformation](./01_linear_transformation.ipynb) in different bases.

**Example:**

If $N$ is diagonal and $P$ is invertible, then $M = P N P^{-1}$ describes the same transformation in the basis formed by the columns of $P$.


```python
import sympy as sp

matrix_n = sp.Matrix([[1, 0], [0, -1]])
matrix_p = sp.Matrix([[1, 1], [1, -1]])
matrix_m = sp.simplify(matrix_p * matrix_n * matrix_p.inv())

print("matrix_m =")
sp.pprint(matrix_m)
```
