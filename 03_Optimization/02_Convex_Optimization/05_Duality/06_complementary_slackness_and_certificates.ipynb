### 3.2.5.6. Complementary Slackness and Certificates


$$\lambda_i^*f_i(x^*)=0$$


**Explanation:**

Complementary slackness pairs each inequality residual with its multiplier. At optimality, a positive multiplier can only sit on an active constraint, and an inactive constraint must have zero multiplier contribution.

The product pairs the optimal multiplier $\lambda_i^*$ with the optimal constraint residual $f_i(x^*)$. It must be zero, so a positive multiplier can occur only when the corresponding inequality is active.

**Properties:**
- Stationarity balances objective and constraint gradients.
- Complementary slackness pairs active constraints with nonzero multipliers.
- Convexity makes the KKT certificate global.

**Intuition:**

The example verifies the nontrivial KKT residuals for one scalar problem.


**Numerical Example:**

For $\min x^2$ subject to $1-x\leq0$, test $x^\star=1$ and $\lambda^\star=2$.

Primal feasibility gives
$$
1-x^\star=1-1=0\leq0.
$$

Dual feasibility gives $\lambda^\star=2\geq0$. Stationarity gives
$$
2x^\star-\lambda^\star=2-2=0.
$$

Complementary slackness gives
$$
\lambda^\star(1-x^\star)=2(1-1)=0.
$$

All KKT residuals vanish.


```python
import sympy as sp

optimizer = sp.Integer(1)
multiplier = sp.Integer(2)
stationarity = 2 * optimizer - multiplier
slackness = multiplier * (1 - optimizer)

print("stationarity =", stationarity)
print("slackness =", slackness)
```
