### 1.1.7.1. Eigenvalue


$$
A \mathbf{e}_{\lambda} = \lambda \mathbf{e}_{\lambda}
$$

$$
p(\lambda) = \det(A - \lambda I) = 0
$$


**Explanation:**

An eigenvalue is the scalar that measures how a matrix stretches a special input direction. It is obtained from the characteristic polynomial and forms the scalar half of an eigenpair, while the corresponding directions appear in [eigenvectors](./02_eigenvector.ipynb).

**Example:**

For

$$
A =
\begin{bmatrix}
9 & -2 \\
-2 & 6
\end{bmatrix},
$$

the characteristic equation gives the eigenvalues $5$ and $10$.


```python
import sympy as sp

matrix_a = sp.Matrix([[9, -2], [-2, 6]])
symbol_lambda = sp.symbols("symbol_lambda")
characteristic_polynomial = sp.expand((matrix_a - symbol_lambda * sp.eye(2)).det())
eigenvalues = sp.solve(sp.Eq(characteristic_polynomial, 0), symbol_lambda)

print("characteristic_polynomial =", characteristic_polynomial)
print("eigenvalues =", eigenvalues)
```
