### 1.4.2.7. Homogeneous Equations

$$
M(tx, ty) = t^{\alpha} M(x, y),\ N(tx, ty) = t^{\alpha} N(x, y)
\;\xrightarrow{\,y = ux,\ dy = u\,dx + x\,du\,}\;
\text{a separable equation in } u, x .
$$

**Explanation:**

A function is **homogeneous of degree $\alpha$** if $f(tx, ty) = t^{\alpha} f(x,y)$. The equation $M\,dx + N\,dy = 0$ is called **homogeneous** when $M$ and $N$ are homogeneous of the *same* degree. The substitution $y = ux$ (so $dy = u\,dx + x\,du$) then reduces it to a [separable equation](./03_separable_equations.ipynb) in the new variable $u = y/x$, which is integrated and back-substituted. (When $M$ is simpler than $N$, the mirror substitution $x = vy$ may be easier.)

**Properties:**
- Degree check: $M(tx,ty)/t^{\alpha} = M(x,y)$ for the common degree $\alpha$.
- After substitution the equation always separates as $\dfrac{dx}{x} + \dfrac{N(1,u)\,du}{M(1,u) + uN(1,u)} = 0$.

**Numerical Example:**

Solve $(x^2 + y^2)\,dx + (x^2 - xy)\,dy = 0$.

**Step 1 — degree.** $M = x^2 + y^2$ and $N = x^2 - xy$ are both homogeneous of degree $2$.

**Step 2 — substitute** $y = ux$, $dy = u\,dx + x\,du$:

$$
(x^2 + u^2 x^2)\,dx + (x^2 - ux^2)(u\,dx + x\,du) = 0 .
$$

**Step 3 — simplify.** Collecting terms (each is $x^2$ or $x^3$),

$$
x^2(1 + u)\,dx + x^3(1 - u)\,du = 0 \;\Rightarrow\; \frac{1 - u}{1 + u}\,du + \frac{dx}{x} = 0 .
$$

**Step 4 — integrate.** Long division gives $\frac{1-u}{1+u} = -1 + \frac{2}{1+u}$, so

$$
\int\!\left(-1 + \frac{2}{1+u}\right)du + \int \frac{dx}{x} = 0
\;\Rightarrow\; -u + 2\ln|1 + u| + \ln|x| = \ln|c| .
$$

**Step 5 — back-substitute** $u = y/x$ and simplify with logarithm laws:

$$
-\frac{y}{x} + 2\ln\Big|1 + \frac{y}{x}\Big| + \ln|x| = \ln|c|
\;\Longrightarrow\; (x + y)^2 = c\,x\,e^{y/x}.
$$

```python
import sympy as sp

x, y, u, t, c = sp.symbols("x y u t c")
M, N = x**2 + y**2, x**2 - x * y

degree_M = sp.simplify(M.subs({x: t * x, y: t * y}) / M)
degree_N = sp.simplify(N.subs({x: t * x, y: t * y}) / N)

separable_integral = sp.integrate((1 - u) / (1 + u), u)

implicit_solution = (x + y)**2 - c * x * sp.exp(y / x)
constant_value = sp.solve(implicit_solution, c)[0]
implicit_derivative = -sp.diff(implicit_solution, x) / sp.diff(implicit_solution, y)
ode_right_side = -M / N
satisfies = sp.simplify(implicit_derivative.subs(c, constant_value) - ode_right_side) == 0

print("M, N homogeneous of degree :", degree_M, degree_N)
print("∫ (1-u)/(1+u) du =", separable_integral)
print("implicit solution (x+y)^2 = c x e^(y/x) satisfies the ODE:", satisfies)
```
