### 1.3.5.1. Functions of Several Variables

$$
f : \mathbb{R}^{n} \to \mathbb{R}, \qquad z = f(x_1, x_2, \ldots, x_n) ;
\qquad
\{\mathbf{x} : f(\mathbf{x}) = c\}\ \text{is a level set}.
$$

**Explanation:**

A function of several variables (a scalar field) assigns one number to each point of an input space $\mathbb{R}^n$. For two inputs its graph is a surface $z = f(x,y)$, and slicing that surface at a fixed height $c$ gives a *level set* (a contour, or level curve), the set of inputs producing the same output. Level sets are how the objective landscapes of optimization, control, and machine learning are visualized — a [cost function](../../../03_Optimization/01_Optimization_Fundamentals/01_mathematical_programming.ipynb) is exactly such a scalar field, and its contours organize the search for a minimum.

**Properties:**
- The domain is the set of inputs where $f$ is defined; the range is the set of attained outputs.
- Level sets of $f(x,y) = x^2 + y^2$ are circles; their nesting shows the bowl rising away from the origin.

**Numerical Example:**

Take $f(x,y) = x^2 + y^2$, a paraboloid opening upward with minimum value $0$ at the origin. Its level sets are

$$
x^2 + y^2 = c \quad\Longrightarrow\quad
\begin{cases}
\text{the single point } (0,0), & c = 0,\\
\text{a circle of radius } \sqrt{c}, & c > 0,\\
\text{empty}, & c < 0.
\end{cases}
$$

For example $f(1,2) = 1 + 4 = 5$, so $(1,2)$ lies on the level curve $x^2 + y^2 = 5$, a circle of radius $\sqrt5 \approx 2.236$ centred at the origin.

```python
import sympy as sp

x, y = sp.symbols("x y")
f = x**2 + y**2

value_at_point = f.subs({x: 1, y: 2})
level_radius = sp.sqrt(value_at_point)

print("f(1, 2)            =", value_at_point)
print("level set x^2+y^2 =", value_at_point, "-> circle of radius", sp.N(level_radius, 6))
```

```python
import numpy as np
import matplotlib.pyplot as plt

grid_axis = np.linspace(-3, 3, 200)
mesh_x, mesh_y = np.meshgrid(grid_axis, grid_axis)
surface = mesh_x**2 + mesh_y**2

fig = plt.figure(figsize=(11, 4))
surface_axis = fig.add_subplot(1, 2, 1, projection="3d")
surface_axis.plot_surface(mesh_x, mesh_y, surface, cmap="viridis", alpha=0.85)
surface_axis.set_title(r"surface $z = x^2 + y^2$")
surface_axis.set_xlabel("x"); surface_axis.set_ylabel("y"); surface_axis.set_zlabel("z")

contour_axis = fig.add_subplot(1, 2, 2)
contour = contour_axis.contour(mesh_x, mesh_y, surface, levels=[1, 5, 9, 13], cmap="viridis")
contour_axis.clabel(contour, inline=True, fontsize=8)
contour_axis.scatter([1], [2], color="red", zorder=5, label="(1, 2) on level 5")
contour_axis.set_title("level sets (contours)")
contour_axis.set_xlabel("x"); contour_axis.set_ylabel("y"); contour_axis.legend(); contour_axis.set_aspect("equal")
plt.tight_layout(); plt.show()
```
