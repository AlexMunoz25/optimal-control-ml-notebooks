### 3.9.14. Damped and Modified Newton Methods

$$
\begin{aligned}
x^{k+1} &= x^k + \alpha^k d^k, \\
\left(\nabla^2 f(x^k)+\Delta^k\right)d^k &= -\nabla f(x^k), \\
d^k &\in \arg\min_{\|d\|\le \gamma^k}
\left\{\nabla f(x^k)^{\mathsf T}d + \frac{1}{2}d^{\mathsf T}\nabla^2f(x^k)d\right\}.
\end{aligned}
$$


**Explanation:**

Modified Newton methods address the main weakness of pure Newton: the Hessian may be singular, indefinite, or simply unreliable far from a minimizer. One remedy is damping, where a stepsize rule is applied to the Newton direction. Another is to modify the Hessian by adding a diagonal correction so that the linear system produces a descent direction. Trust-region methods provide another safeguard by restricting the step to a neighborhood where the quadratic model is expected to represent the true cost reasonably well.

The first line is a damped step with direction $d^k$ and stepsize $\alpha^k$. The second line solves a modified Newton system with Hessian correction $\Delta^k$, and the third line reads the trust-region step as the minimizer of a local quadratic model inside a [norm](../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb) ball.

The guiding idea is to preserve Newton's fast local behavior without trusting it blindly everywhere. Near a nonsingular local minimum, the correction should disappear and the method should reduce to pure Newton. Away from such a point, safeguards based on [Cholesky factorization](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/27_cholesky_decomposition.ipynb), [positive definite](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/17_positive_definite_matrix.ipynb) corrections, or trust regions keep the iteration meaningful as a minimization method.

**Intuition:**

If the Hessian has a negative [eigenvalue](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/08_eigenvalue.ipynb), the pure Newton direction may not decrease the cost. Adding a diagonal correction can make the model Hessian positive definite, and solving the corrected system gives a descent direction. The code cell demonstrates this idea on a small indefinite Hessian by checking that the corrected direction has negative directional derivative.


**Numerical Example:**

Let the Hessian approximation and gradient be
$$
H=\begin{bmatrix}-2&0\\0&4\end{bmatrix},
\qquad g=\begin{bmatrix}1\\-3\end{bmatrix}.
$$

The Hessian has eigenvalues $-2$ and $4$, so it is indefinite. Add a diagonal correction
$$
\Delta=3I=\begin{bmatrix}3&0\\0&3\end{bmatrix},
\qquad H+\Delta=\begin{bmatrix}1&0\\0&7\end{bmatrix}.
$$

Now the corrected Newton equation is
$$
(H+\Delta)d=-g
\quad\Longrightarrow\quad
\begin{bmatrix}1&0\\0&7\end{bmatrix}
\begin{bmatrix}d_1\\d_2\end{bmatrix}
=
\begin{bmatrix}-1\\3\end{bmatrix}.
$$

Thus
$$
d_1=-1,
\qquad 7d_2=3,
\qquad d_2=\frac{3}{7}.
$$

The directional derivative is
$$
g^{\mathsf T}d=1(-1)+(-3)\left(\frac{3}{7}\right)
=-1-\frac{9}{7}
=-\frac{16}{7}<0,
$$
so the corrected direction is a descent direction.


```python
import sympy as sp

hessian = sp.Matrix([[-2, 0], [0, 4]])
gradient = sp.Matrix([1, -3])
diagonal_correction = 3 * sp.eye(2)
modified_direction = (hessian + diagonal_correction).solve(-gradient)
descent_check = (gradient.T * modified_direction)[0]

print("modified_direction ="); sp.pprint(modified_direction)
print("descent_check =", descent_check)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

hessian = ca.DM([[-2, 0], [0, 4]])
gradient = ca.DM([1, -3])
diagonal_correction = 3 * ca.DM.eye(2)
modified_direction = ca.solve(hessian + diagonal_correction, -gradient)
descent_check = ca.dot(gradient, modified_direction)

print("modified_direction =", modified_direction)
print("descent_check =", descent_check)
```
