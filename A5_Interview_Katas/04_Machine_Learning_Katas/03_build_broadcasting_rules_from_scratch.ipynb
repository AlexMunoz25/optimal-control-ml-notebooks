### A5.4.3. Build Broadcasting Rules from Scratch

**Problem:**

Build a function that, given two tensor shapes, determines whether they are broadcast-compatible and computes the resulting broadcast shape.

**Explanation:**

Broadcasting allows operations between tensors of different shapes by virtually expanding the smaller one. The rules are applied dimension by dimension, right to left.

How to build it:

1. Pad the shorter shape on the left with ones so both shapes have the same number of dimensions.
2. Walk through dimensions from right to left.
3. For each pair of dimensions: if they are equal, the output dimension is that value. If one of them is 1, the output dimension is the other value. If neither condition holds, the shapes are incompatible.
4. Collect all output dimensions to form the broadcast shape.

**Example:**

Shapes `(3, 1)` and `(1, 4)` broadcast to `(3, 4)`. Shape `(5, 3)` and `(4, 3)` are incompatible because 5 ≠ 4 and neither is 1.

```python
def broadcast_shapes(shape_a, shape_b):
    max_dims = max(len(shape_a), len(shape_b))
    padded_a = (1,) * (max_dims - len(shape_a)) + tuple(shape_a)
    padded_b = (1,) * (max_dims - len(shape_b)) + tuple(shape_b)

    result = []
    for dim_a, dim_b in zip(padded_a, padded_b):
        if dim_a == dim_b:
            result.append(dim_a)
        elif dim_a == 1:
            result.append(dim_b)
        elif dim_b == 1:
            result.append(dim_a)
        else:
            return None

    return tuple(result)


test_cases = [
    ((3, 1), (1, 4)),
    ((2, 3, 4), (4,)),
    ((5, 1, 3), (1, 7, 3)),
    ((2, 3), (3,)),
    ((5, 3), (4, 3)),
]

for shape_a, shape_b in test_cases:
    result = broadcast_shapes(shape_a, shape_b)
    status = f"→ {result}" if result else "→ incompatible"
    print(f"{shape_a} ⊕ {shape_b} {status}")
```
