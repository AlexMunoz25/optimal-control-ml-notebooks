### 3.1.25. Quasi-Newton Scaling and Comparison


$$
\tilde D^0
= \frac{p^{0\mathsf T}q^0}{q^{0\mathsf T}D^0q^0}D^0,
\qquad
R^k = Q^{1/2}D^kQ^{1/2}.
$$


**Explanation:**

The initial scaling of a quasi-Newton matrix can strongly influence the early iterations, before the method has accumulated enough curvature pairs. Bertsekas discusses scaling because multiplying $D^0$ by a suitable positive scalar can make the first approximate inverse Hessian better match the observed curvature. The condition number of $R^k$ describes the scaled local geometry, connecting quasi-Newton scaling back to [spectral properties](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/30_spectral_radius.ipynb), [eigendecomposition](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/12_eigendecomposition.ipynb), and [condition numbers](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/31_condition_number.ipynb).

The comparison with conjugate gradient is also practical. Quasi-Newton methods tend to be strong when gradient evaluations are expensive enough that the extra matrix storage and matrix-vector multiplication are acceptable. Conjugate gradient is often preferred in very large problems where storing a full matrix is too costly and matrix-vector products are cheap.

**Intuition:**

A scalar rescaling of $D^0$ uses the first step and gradient change to match curvature along the observed direction. The Bertsekas figure relates this scaling idea to the condition number of a quasi-Newton iteration. The code cell computes the simple scalar scaling factor from one curvature pair.

<img src="../../Figures/030101_berts_fig_1_7_5_quasi_newton_condition_number.jpeg" alt="Bertsekas Figure 1.7.5: condition number of a quasi-Newton iteration" style="width: 60%; height: auto;">


**Numerical Example:**

The condition-number figure motivates scaling the initial inverse-Hessian approximation. Let
$$
D^0=I,
\qquad p^0=\begin{bmatrix}1\\2\end{bmatrix},
\qquad q^0=\begin{bmatrix}3\\5\end{bmatrix}.
$$

The numerator is
$$
p^{0\mathsf T}q^0=1\cdot 3+2\cdot 5=13.
$$

The denominator is
$$
q^{0\mathsf T}D^0q^0=q^{0\mathsf T}q^0=3^2+5^2=34.
$$

Therefore the scalar scaling factor is
$$
\frac{p^{0\mathsf T}q^0}{q^{0\mathsf T}D^0q^0}=\frac{13}{34}\approx 0.3824.
$$

The scaled starting matrix is
$$
\tilde{D}^0=0.3824I
=\begin{bmatrix}0.3824&0\\0&0.3824\end{bmatrix}.
$$


```python
import numpy as np

initial_inverse_hessian = np.eye(2)
position_increment = np.array([1.0, 2.0])
gradient_increment = np.array([3.0, 5.0])
scale_factor = (position_increment @ gradient_increment) / (gradient_increment @ initial_inverse_hessian @ gradient_increment)
scaled_inverse_hessian = scale_factor * initial_inverse_hessian

scale_factor, scaled_inverse_hessian
```
