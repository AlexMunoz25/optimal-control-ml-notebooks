### 3.2.2.3. Cones and Conic Combinations


$$\theta_1x_1+\cdots+\theta_kx_k,\qquad\theta_i\geq0$$


**Explanation:**

A convex cone is closed under nonnegative linear combinations. This is the set geometry behind conic constraints and generalized inequalities. Feasibility means the combination remains inside the cone; optimality conditions later use dual cone multipliers to certify such constraints.

**Properties:**
- Conic coefficients are nonnegative.
- They do not need to sum to one.
- The zero vector appears when all coefficients are zero.

**Intuition:**

The figure above shows the cone generated by rays; the arithmetic computes one nonnegative combination inside it.

**Numerical Example:**

The figure above shows the cone generated by rays; the arithmetic computes one nonnegative combination inside it.

Let
$$
v_1=\begin{bmatrix}1\\0\end{bmatrix},\quad
v_2=\begin{bmatrix}1\\2\end{bmatrix}.
$$

Choose nonnegative coefficients $3$ and $2$. Then
$$
3v_1+2v_2=3\begin{bmatrix}1\\0\end{bmatrix}+2\begin{bmatrix}1\\2\end{bmatrix}.
$$

The coordinates are
$$
3\cdot1+2\cdot1=5,
\qquad
3\cdot0+2\cdot2=4.
$$

So the conic combination is
$$
\begin{bmatrix}5\\4\end{bmatrix}.
$$


```python
import sympy as sp

direction_1 = sp.Matrix([1, 0])
direction_2 = sp.Matrix([1, 2])
combination = 3 * direction_1 + 2 * direction_2

combination
```
