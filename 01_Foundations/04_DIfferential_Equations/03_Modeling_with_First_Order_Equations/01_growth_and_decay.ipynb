### 1.4.3.1. Growth and Decay

$$
\frac{dA}{dt} = kA, \quad A(0) = A_0 \;\Longrightarrow\; A(t) = A_0 e^{kt} .
$$

**Explanation:**

When a quantity changes at a rate proportional to its current size, the model is $A' = kA$, whose [separable / linear](../02_First_Order_Equations/03_separable_equations.ipynb) solution is the exponential $A(t) = A_0 e^{kt}$. The sign of the constant decides the behaviour: $k > 0$ is a **growth constant** (populations, bacteria, compound interest) and $k < 0$ is a **decay constant** (radioactive disintegration). For decay, the **half-life** is the time for half of an initial amount to disappear; it depends only on $k$, not on $A_0$.

**Numerical Example:**

**Growth.** A bacterial culture has $P_0$ cells, and $P(1) = \tfrac{3}{2}P_0$. With $\dfrac{dP}{dt} = kP$, $P(0) = P_0$ the solution is $P(t) = P_0 e^{kt}$. Use the measurement at $t = 1$:

$$
\tfrac{3}{2}P_0 = P_0 e^{k} \;\Rightarrow\; e^{k} = \tfrac{3}{2} \;\Rightarrow\; k = \ln\tfrac{3}{2} \approx 0.4055 .
$$

Time to **triple**: solve $3P_0 = P_0 e^{kt}$, i.e. $kt = \ln 3$, so

$$
t = \frac{\ln 3}{0.4055} \approx 2.71 \text{ hours}.
$$

**Decay (half-life).** Plutonium-239 with $A(t) = A_0 e^{kt}$ loses $0.043\%$ in 15 years, so $A(15) = 0.99957\,A_0$:

$$
0.99957\,A_0 = A_0 e^{15k} \;\Rightarrow\; k = \tfrac{1}{15}\ln 0.99957 \approx -2.867\times 10^{-5} .
$$

The half-life solves $\tfrac12 A_0 = A_0 e^{kt}$, i.e. $kt = -\ln 2$:

$$
t = \frac{\ln 2}{2.867\times 10^{-5}} \approx 24{,}180 \text{ years}.
$$

```python
import sympy as sp

t, k = sp.symbols("t k")
P_0 = sp.symbols("P_0", positive=True)
P = sp.Function("P")

solution = sp.dsolve(sp.Eq(P(t).diff(t), k * P(t)), P(t), ics={P(0): P_0})
growth_constant = sp.solve(sp.Eq(solution.rhs.subs(t, 1), sp.Rational(3, 2) * P_0), k)[0]
triple_time = sp.solve(sp.Eq(P_0 * sp.exp(growth_constant * t), 3 * P_0), t)[0]

decay_constant = sp.Rational(1, 15) * sp.log(sp.Rational(99957, 100000))
half_life = sp.solve(sp.Eq(sp.exp(decay_constant * t), sp.Rational(1, 2)), t)[0]

print("growth: P(t) =", solution.rhs)
print("  k = ln(3/2)     =", sp.N(growth_constant, 5))
print("  triple time     =", sp.N(triple_time, 4), "hours")
print("decay: k          =", sp.N(decay_constant, 4))
print("  half-life       =", sp.N(half_life, 5), "years")
```
