### 1.3.1.2. Limit Laws

$$
\begin{aligned}
\lim_{x \to a}\big(f(x) \pm g(x)\big) &= L \pm M, &
\lim_{x \to a}\big(c\,f(x)\big) &= c\,L, \\
\lim_{x \to a}\big(f(x)\,g(x)\big) &= L\,M, &
\lim_{x \to a}\frac{f(x)}{g(x)} &= \frac{L}{M}\ (M \neq 0), \\
\lim_{x \to a}\big(f(x)\big)^{n} &= L^{n}, &
\lim_{x \to a}\sqrt[n]{f(x)} &= \sqrt[n]{L}.
\end{aligned}
$$

**Explanation:**

When $\lim_{x\to a} f(x) = L$ and $\lim_{x\to a} g(x) = M$ both exist, the limit operation distributes over sums, differences, constant multiples, products, quotients (provided $M \neq 0$), powers, and roots. Together with the two basic limits $\lim_{x\to a} c = c$ and $\lim_{x\to a} x = a$, these laws let every polynomial be evaluated by direct substitution, and every rational function wherever its denominator is nonzero. They reduce limit computation from tables and graphs to ordinary algebra.

**Assumptions:**
- $L = \lim_{x\to a} f(x)$ and $M = \lim_{x\to a} g(x)$ each exist (are finite).
- For the quotient law, $M \neq 0$; for the even root law, $L \geq 0$.

**Numerical Example:**

Evaluate $\lim_{x \to -3}(4x + 2)$ one law at a time:

$$
\lim_{x \to -3}(4x + 2) = \lim_{x \to -3} 4x + \lim_{x \to -3} 2 \quad\text{(sum law)}
$$

$$
= 4 \cdot \lim_{x \to -3} x + \lim_{x \to -3} 2 \quad\text{(constant multiple law)}
$$

$$
= 4 \cdot (-3) + 2 = -10 \quad\text{(basic limits)}.
$$

For a quotient, $\lim_{x \to 2}\dfrac{x^2 + 1}{x - 3} = \dfrac{\lim_{x\to 2}(x^2+1)}{\lim_{x\to 2}(x-3)} = \dfrac{5}{-1} = -5$, valid because the denominator limit $-1 \neq 0$.

```python
import sympy as sp

x = sp.symbols("x")
c = sp.Integer(4)

linear_expression = 4 * x + 2
sum_split = sp.limit(4 * x, x, -3) + sp.limit(sp.Integer(2), x, -3)
constant_multiple_step = c * sp.limit(x, x, -3) + 2
linear_limit = sp.limit(linear_expression, x, -3)

quotient_expression = (x**2 + 1) / (x - 3)
numerator_limit = sp.limit(x**2 + 1, x, 2)
denominator_limit = sp.limit(x - 3, x, 2)
quotient_limit = sp.limit(quotient_expression, x, 2)

print("sum law      :", sum_split)
print("const-mult   :", constant_multiple_step)
print("lim (4x+2)   :", linear_limit)
print("numerator    :", numerator_limit, " denominator :", denominator_limit)
print("lim quotient :", quotient_limit)
```
