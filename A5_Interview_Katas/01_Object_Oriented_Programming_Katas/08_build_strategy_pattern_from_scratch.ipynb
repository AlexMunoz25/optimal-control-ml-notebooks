### A5.1.8. Build Strategy Pattern from Scratch

**Problem:**

Build a class that delegates an algorithm to a swappable strategy object. The strategy can be changed at runtime without modifying the class itself.

**Explanation:**

The strategy pattern separates an algorithm from the object that uses it. The object holds a reference to a strategy, and delegates work to it. Swapping the strategy changes the behavior without changing the object's code.

How to build it:

1. Define a common interface for strategies. In Python, this means each strategy has the same method name (e.g., `execute`).
2. Create concrete strategy classes, each implementing `execute` differently.
3. Create a context class that stores a strategy in `__init__` and calls `self.strategy.execute(...)` when it needs to perform the operation.
4. Provide a `set_strategy(strategy)` method to swap the strategy at runtime.

**Example:**

A `Sorter` context holds a sorting strategy. Swap between `BubbleSortStrategy` and `QuickSortStrategy` to change how data is sorted.

```python
class BubbleSortStrategy:
    def execute(self, data):
        result = list(data)
        length = len(result)
        for outer in range(length):
            for inner in range(length - outer - 1):
                if result[inner] > result[inner + 1]:
                    result[inner], result[inner + 1] = result[inner + 1], result[inner]
        return result


class BuiltinSortStrategy:
    def execute(self, data):
        return sorted(data)


class Sorter:
    def __init__(self, strategy):
        self.strategy = strategy

    def set_strategy(self, strategy):
        self.strategy = strategy

    def sort(self, data):
        return self.strategy.execute(data)


data = [5, 3, 8, 1, 9, 2]

sorter = Sorter(BubbleSortStrategy())
print(f"Bubble sort: {sorter.sort(data)}")

sorter.set_strategy(BuiltinSortStrategy())
print(f"Builtin sort: {sorter.sort(data)}")
```
