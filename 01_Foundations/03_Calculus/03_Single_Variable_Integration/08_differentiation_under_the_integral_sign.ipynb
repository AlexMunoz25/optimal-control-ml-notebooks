### 1.3.3.8. Differentiation Under the Integral Sign

$$
\frac{d}{dx}\int_{a(x)}^{b(x)} f(t)\,dt = f\big(b(x)\big)b'(x) - f\big(a(x)\big)a'(x),
$$

$$
\frac{\partial}{\partial \theta}\int_{a}^{b} f(x, \theta)\,dx = \int_{a}^{b} \frac{\partial f}{\partial \theta}(x, \theta)\,dx \quad (\text{Leibniz rule}).
$$

**Explanation:**

There are two ways an integral can depend on a variable, and each has a differentiation rule. When the *limits* depend on $x$, the [fundamental theorem](./03_fundamental_theorem_of_calculus.ipynb) and [chain rule](../02_Single_Variable_Differentiation/03_chain_rule.ipynb) give the boundary-evaluation form. When the *integrand* depends on a parameter $\theta$, Leibniz's rule lets the derivative pass inside the integral and act on the integrand alone. The parameter form is the backbone of the [score-function / policy gradient](../../../07_Reinforcement_Learning) and reparameterization estimators, where a gradient of an expectation is rewritten as an expectation of a gradient.

**Properties:**
- Validity of swapping $\tfrac{d}{d\theta}$ and $\int$ requires $f$ and $\partial_\theta f$ to be continuous (dominated-convergence conditions).
- The general (full) Leibniz rule combines both effects when limits and integrand all depend on the parameter.

**Numerical Example:**

Variable-limit form: let $G(x) = \int_{0}^{x^{2}} \sin t\,dt$. With $b(x) = x^2$, $b'(x) = 2x$,

$$
G'(x) = \sin(x^2)\cdot 2x = 2x\sin(x^2) .
$$

Parameter form: let $I(\theta) = \int_{0}^{1} e^{\theta x}\,dx = \dfrac{e^{\theta} - 1}{\theta}$. Differentiating under the integral,

$$
I'(\theta) = \int_{0}^{1} \frac{\partial}{\partial\theta} e^{\theta x}\,dx = \int_{0}^{1} x\,e^{\theta x}\,dx ,
$$

and at $\theta = 1$ this equals $\int_0^1 x e^{x}\,dx = 1$, which matches differentiating the closed form $\tfrac{e^\theta-1}{\theta}$ directly.

```python
import sympy as sp

x, t, theta = sp.symbols("x t theta")

variable_limit_integral = sp.integrate(sp.sin(t), (t, 0, x**2))
variable_limit_derivative = sp.diff(variable_limit_integral, x)

closed_form = (sp.exp(theta) - 1) / theta
closed_form_derivative = sp.simplify(sp.diff(closed_form, theta))

leibniz_inside = sp.integrate((x * sp.exp(theta * x)).subs(theta, 1), (x, 0, 1))
closed_form_derivative_at_one = closed_form_derivative.subs(theta, 1)

print("G(x) = ∫_0^{x^2} sin t dt =", variable_limit_integral)
print("G'(x)                      =", sp.simplify(variable_limit_derivative))
print("I(θ) = (e^θ - 1)/θ,  I'(θ) =", closed_form_derivative)
print("Leibniz swap: ∫_0^1 x e^x dx =", leibniz_inside)
print("closed-form I'(1)            =", closed_form_derivative_at_one)
```

**Equivalent CasADi implementation:**

Building $I(\theta)$ from an analytic inner integral and differentiating with `ca.gradient` reproduces $I'(1)$, the gradient-of-an-integral pattern AD performs automatically.

```python
import casadi as ca

theta = ca.SX.sym("theta")
closed_form = (ca.exp(theta) - 1) / theta
derivative_function = ca.Function("dI", [theta], [ca.gradient(closed_form, theta)])

print("I'(1) via CasADi =", float(derivative_function(1.0)))
```
