### 3.3.1. Equality Constraint Lagrange Multiplier Theorem

$$
\begin{aligned}
\min_{x \in \mathbb{R}^n} \quad & f(x) \\
\text{subject to} \quad & h_i(x)=0, \quad i=1,\ldots,m,
\end{aligned}
$$

$$
\nabla f(x^*) + \sum_{i=1}^{m} \lambda_i^* \nabla h_i(x^*) = 0
$$

**Explanation:**

The [vector](../../01_Foundations/01_Linear_Algebra/02_Vector/01_vector_definition.ipynb) $x$ is the decision point, $f$ is the cost, and $h_i(x)=0$ are equality constraints that define the feasible surface. At a regular local minimum, the cost gradient cannot point along a first-order feasible variation, so it must be built from the constraint gradients. The multipliers $\lambda_i^*$ are the coefficients in that linear combination; they turn the constrained first-order condition into the stationarity condition of the Lagrangian.

**Assumptions:**
- The functions $f$ and $h_i$ are continuously differentiable near $x^*$.
- The constraint gradients $\nabla h_1(x^*),\ldots,\nabla h_m(x^*)$ are linearly independent and [span](../../01_Foundations/01_Linear_Algebra/05_Geometrical_Aspects_of_Linear_Algebra/14_span.ipynb) the local normal directions.

**Intuition:**

<img src="../../Figures/03030301_berts_fig_3_1_1_lagrange_multiplier_condition.jpeg" alt="Bertsekas Figure 3.1.1: cost gradient collinear with the constraint gradient at an equality-constrained minimum" style="width: 60%; height: auto;">

The figure shows a linear cost minimized on a circle. At the best feasible point, the cost gradient and the circle's normal direction are collinear, so the gradient of the objective is canceled by a multiple of the constraint gradient.

**Numerical Example:**

Use the same circle example shown in the figure:

$$
\min_{x_1,x_2} x_1+x_2 \quad \text{subject to} \quad h(x)=x_1^2+x_2^2-2=0.
$$

The feasible point highlighted in the figure is

$$
x^*=\begin{bmatrix}-1\\-1\end{bmatrix}, \qquad h(x^*)=(-1)^2+(-1)^2-2=0.
$$

The objective gradient and constraint gradient are

$$
\nabla f(x^*)=\begin{bmatrix}1\\1\end{bmatrix},
\qquad
\nabla h(x^*)=\begin{bmatrix}2x_1^*\\2x_2^*\end{bmatrix}
=\begin{bmatrix}-2\\-2\end{bmatrix}.
$$

The multiplier equation is

$$
\begin{bmatrix}1\\1\end{bmatrix}
+\lambda
\begin{bmatrix}-2\\-2\end{bmatrix}
=\begin{bmatrix}0\\0\end{bmatrix}.
$$

Both coordinates give

$$
1-2\lambda=0 \quad \Rightarrow \quad \lambda=\frac{1}{2}.
$$

Therefore the objective gradient is exactly canceled by one half of the constraint gradient, matching the normal-direction geometry in the figure.

```python
import sympy as sp

x_1, x_2, lambda_multiplier = sp.symbols('x_1 x_2 lambda')

cost = x_1 + x_2
constraint = x_1**2 + x_2**2 - 2
candidate = {x_1: -1, x_2: -1}

gradient_cost = sp.Matrix([sp.diff(cost, x_1), sp.diff(cost, x_2)]).subs(candidate)
gradient_constraint = sp.Matrix([sp.diff(constraint, x_1), sp.diff(constraint, x_2)]).subs(candidate)
stationarity_equation = gradient_cost + lambda_multiplier * gradient_constraint
lambda_solution = sp.solve(stationarity_equation, [lambda_multiplier], dict=True)

result = {
    'gradient_cost': gradient_cost,
    'gradient_constraint': gradient_constraint,
    'lambda_solution': lambda_solution,
    'stationarity_at_solution': stationarity_equation.subs(lambda_solution[0]),
}

for label, value in result.items():
    print(f"{label} =", value)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

opti = ca.Opti()
decision_variable = opti.variable(2)
opti.minimize(decision_variable[0] + decision_variable[1])
opti.subject_to(decision_variable[0] ** 2 + decision_variable[1] ** 2 == 2)
opti.set_initial(decision_variable, [-1, -1])
opti.solver("ipopt", {"print_time": False}, {"print_level": 0, "sb": "yes"})
solution = opti.solve()

print("minimizer =", ca.DM(solution.value(decision_variable)))
```
