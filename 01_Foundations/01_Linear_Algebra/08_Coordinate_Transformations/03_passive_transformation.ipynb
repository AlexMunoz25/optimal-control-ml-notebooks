### 1.1.8.3. Passive Transformation


$$
\begin{aligned}
\begin{pmatrix}
\text{Components of} \\
\text{same vector} \\
\text{in new system}
\end{pmatrix}
&=
\begin{pmatrix}
\text{Inverse} \\
\text{transformation} \\
\text{matrix}
\end{pmatrix}
\begin{pmatrix}
\text{Components of} \\
\text{vector in} \\
\text{original system}
\end{pmatrix}, \\
R_{\mathrm{passive}}(\theta)
&=
\begin{bmatrix}
\cos \theta & \sin \theta \\
-\sin \theta & \cos \theta
\end{bmatrix}
=
R_{\mathrm{direct}}(\theta)^{-1}
\end{aligned}
$$


**Explanation:**

A passive transformation changes the coordinate system while the geometric vector remains fixed. The basis vectors move by the direct transformation matrix, so the component column must move by the inverse matrix to preserve the same vector. For orthogonal rotations, the inverse is the transpose, which makes the passive and direct matrices easy to compare.

**Example:**

For the same $150^{\circ}$ angle,
$$
R_{\mathrm{passive}}(150^{\circ})
R_{\mathrm{direct}}(150^{\circ})
=
I,
$$
so the component law for the unchanged vector is the inverse of the basis-vector law.


```python
import numpy as np

rotation_angle_degrees = 150.0
rotation_angle = np.deg2rad(rotation_angle_degrees)
passive_transformation_matrix = np.array(
    [
        [np.cos(rotation_angle), np.sin(rotation_angle)],
        [-np.sin(rotation_angle), np.cos(rotation_angle)],
    ]
)
direct_transformation_matrix = np.array(
    [
        [np.cos(rotation_angle), -np.sin(rotation_angle)],
        [np.sin(rotation_angle), np.cos(rotation_angle)],
    ]
)
original_vector_components = np.array([5.0, 3.0])
transformed_vector_components = passive_transformation_matrix @ original_vector_components
matrix_product = passive_transformation_matrix @ direct_transformation_matrix

print("transformed_vector_components =", np.round(transformed_vector_components, 4))
print("passive_transformation_matrix @ direct_transformation_matrix =")
print(np.round(matrix_product, 4))
```
