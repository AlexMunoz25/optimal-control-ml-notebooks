### 1.4.8.4. Matrix Exponential

$$
e^{At} = I + At + A^2\frac{t^2}{2!} + \cdots = \sum_{k=0}^{\infty} A^k\frac{t^k}{k!},
\qquad
\mathbf{X}(t) = e^{At}\,\mathbf{X}(0) \ \text{ solves } \ \mathbf{X}' = A\mathbf{X} .
$$

**Explanation:**

The **matrix exponential** $e^{At}$ generalizes the scalar $e^{at}$ by the same power series with $A$ in place of $a$. It is the single solution operator of the linear system: $\mathbf{X}(t) = e^{At}\mathbf{X}(0)$ solves $\mathbf{X}' = A\mathbf{X}$ for any initial state, since $\tfrac{d}{dt}e^{At} = Ae^{At}$ and $e^{A\cdot 0} = I$. It is the **state-transition matrix** of a linear time-invariant system in control, and via the [eigendecomposition](../../01_Linear_Algebra/07_Theoretical_Linear_Algebra/12_eigendecomposition.ipynb) $A = K\Lambda K^{-1}$ it factors as $e^{At} = K e^{\Lambda t} K^{-1}$, tying it back to the [eigenvalue method](./02_homogeneous_systems_eigenvalue_method.ipynb).

**Properties:**
- For a diagonal $A = \operatorname{diag}(a_1, \ldots, a_n)$, $e^{At} = \operatorname{diag}(e^{a_1 t}, \ldots, e^{a_n t})$.
- $e^{A(t+s)} = e^{At}e^{As}$ and $\big(e^{At}\big)^{-1} = e^{-At}$ (the semigroup / state-transition properties).

**Numerical Example:**

**Diagonal case.** For $A = \begin{pmatrix} 2 & 0 \\ 0 & 3 \end{pmatrix}$, the powers are $A^k = \begin{pmatrix} 2^k & 0 \\ 0 & 3^k \end{pmatrix}$, so the series splits row by row:

$$
e^{At} = \begin{pmatrix} \sum_k \frac{(2t)^k}{k!} & 0 \\ 0 & \sum_k \frac{(3t)^k}{k!} \end{pmatrix} = \begin{pmatrix} e^{2t} & 0 \\ 0 & e^{3t} \end{pmatrix}.
$$

**General case.** For $A = \begin{pmatrix} 2 & 3 \\ 2 & 1 \end{pmatrix}$ (eigenvalues $-1, 4$ from the eigenvalue notebook), summing the series gives

$$
e^{At} = \frac{1}{5}\begin{pmatrix} 3e^{4t} + 2e^{-t} & 3e^{4t} - 3e^{-t} \\ 2e^{4t} - 2e^{-t} & 2e^{4t} + 3e^{-t} \end{pmatrix},
$$

and the solution from state $\mathbf{X}(0) = (1, 0)^\top$ is $\mathbf{X}(t) = e^{At}\mathbf{X}(0) = \tfrac15\big(3e^{4t} + 2e^{-t},\ 2e^{4t} - 2e^{-t}\big)^\top$ — the same combination of $e^{-t}$ and $e^{4t}$ modes found by the eigenvalue method.

```python
import sympy as sp

t = sp.symbols("t")

diagonal = sp.Matrix([[2, 0], [0, 3]])
general = sp.Matrix([[2, 3], [2, 1]])

exp_diagonal = (diagonal * t).exp()
exp_general = sp.simplify((general * t).exp())
derivative_check = sp.simplify(sp.diff(exp_general, t) - general * exp_general)
state = exp_general * sp.Matrix([1, 0])

print("e^{At}, A diagonal ="); sp.pprint(exp_diagonal)
print("e^{At}, A general  ="); sp.pprint(exp_general)
print("d/dt e^{At} - A e^{At} =", derivative_check.T, "(zero -> solves X' = AX)")
print("X(t) = e^{At} X(0), X(0)=(1,0):"); sp.pprint(sp.simplify(state))
```
