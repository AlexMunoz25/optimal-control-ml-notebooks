### 1.1.8.1. Coordinate-System Transformation


$$
\begin{bmatrix}
A^{x'} \\
A^{y'}
\end{bmatrix}
=
\begin{bmatrix}
\cos \theta & \sin \theta \\
-\sin \theta & \cos \theta
\end{bmatrix}
\begin{bmatrix}
A^x \\
A^y
\end{bmatrix}
$$


**Explanation:**

Fleisch begins Chapter 4 with the change in vector components caused by rotating the coordinate axes. The geometric vector itself does not change, but its components do because the reference directions change. This notebook treats that coordinate-system transformation as the foundational concept for the rest of the module.

  <img src="../../../Figures/01010801_rotation_2d_coordinate_system.jpeg" alt="Rotation of a two-dimensional coordinate system" width="420">
</p>

**Example:**

If a vector has components $(5, 3)$ and the axes rotate by $150^{\circ}$, the new components are obtained by multiplying the old component column by the coordinate-transformation matrix.


```python
import numpy as np

rotation_angle = np.deg2rad(150.0)
coordinate_transformation = np.array(
    [
        [np.cos(rotation_angle), np.sin(rotation_angle)],
        [-np.sin(rotation_angle), np.cos(rotation_angle)],
    ]
)
vector_components = np.array([5.0, 3.0])
rotated_components = coordinate_transformation @ vector_components

print("rotated_components =", np.round(rotated_components, 4))
```
