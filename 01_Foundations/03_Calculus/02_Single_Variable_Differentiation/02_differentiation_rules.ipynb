### 1.3.2.2. Differentiation Rules

$$
\frac{d}{dx}\big(c\big) = 0, \qquad
\frac{d}{dx}\big(x^{n}\big) = n x^{n-1}, \qquad
\frac{d}{dx}\big(f \pm g\big) = f' \pm g',
$$

$$
\frac{d}{dx}\big(f g\big) = f' g + f g', \qquad
\frac{d}{dx}\!\left(\frac{f}{g}\right) = \frac{f' g - f g'}{g^{2}}\ \ (g \neq 0).
$$

**Explanation:**

These rules turn the limit definition into algebra, letting any polynomial, rational, or product of differentiable functions be differentiated without returning to the difference quotient. The constant, power, sum, and constant-multiple rules handle polynomials term by term. The product rule does **not** equal $f'g'$: the derivative of a product is "derivative of the first times the second, plus the first times the derivative of the second," and the quotient rule extends this to ratios. Together with the [chain rule](./03_chain_rule.ipynb) they cover essentially every elementary derivative computed downstream.

**Properties:**
- Linearity: $\frac{d}{dx}(\alpha f + \beta g) = \alpha f' + \beta g'$ for constants $\alpha, \beta$.
- The quotient rule follows from the product rule applied to $f \cdot g^{-1}$.

**Numerical Example:**

Differentiate the product $j(x) = (x^2 + 1)(3x - 2)$ with the product rule. Here $f = x^2 + 1$, $g = 3x - 2$, so $f' = 2x$, $g' = 3$:

$$
j'(x) = f'g + fg' = (2x)(3x - 2) + (x^2 + 1)(3) = 6x^2 - 4x + 3x^2 + 3 = 9x^2 - 4x + 3 .
$$

For the quotient $q(x) = \dfrac{x^2 + 1}{3x - 2}$,

$$
q'(x) = \frac{f'g - fg'}{g^2} = \frac{(2x)(3x - 2) - (x^2 + 1)(3)}{(3x - 2)^2}
= \frac{3x^2 - 4x - 3}{(3x - 2)^2} .
$$

```python
import sympy as sp

x = sp.symbols("x")
f = x**2 + 1
g = 3 * x - 2

product_rule = sp.diff(f, x) * g + f * sp.diff(g, x)
product_check = sp.diff(f * g, x)
quotient_rule = (sp.diff(f, x) * g - f * sp.diff(g, x)) / g**2
quotient_check = sp.diff(f / g, x)

print("product rule    :", sp.expand(product_rule))
print("sympy d/dx(fg)  :", sp.expand(product_check))
print("quotient rule   :", sp.simplify(quotient_rule))
print("sympy d/dx(f/g) :", sp.simplify(quotient_check))
```

**Equivalent CasADi implementation:**

CasADi produces the same derivatives directly from the product and quotient expressions.

```python
import casadi as ca

x = ca.SX.sym("x")
f = x**2 + 1
g = 3 * x - 2

product_derivative = ca.gradient(f * g, x)
quotient_derivative = ca.gradient(f / g, x)

print("d/dx (fg)  =", ca.simplify(product_derivative))
print("d/dx (f/g) =", ca.simplify(quotient_derivative))
```
