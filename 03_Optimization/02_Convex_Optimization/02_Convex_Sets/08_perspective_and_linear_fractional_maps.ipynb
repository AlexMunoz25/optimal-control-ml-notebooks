### 3.2.2.8. Perspective and Linear-Fractional Maps


$$P(z,t)=z/t,\qquad t>0$$


**Explanation:**

The perspective map normalizes coordinates by a positive scale. Linear-fractional maps combine an affine transformation with this normalization, which explains why many ratio expressions become convex-set transformations after lifting.

**Properties:**
- The scaling coordinate must be positive.
- The map sends rays to normalized coordinates.
- Linear-fractional maps combine affine maps with this normalization.

**Intuition:**

The figure above gives the geometric normalization; the arithmetic performs the same normalization numerically.

<img src="../../../Figures/03020208_bv_fig_2_15_perspective_map.jpeg" alt="Boyd and Vandenberghe Figure 2.15: pin-hole camera interpretation of the perspective function" style="width: 60%; height: auto;">


**Numerical Example:**

The figure above gives the geometric normalization; the arithmetic performs the same normalization numerically.

Let
$$
z=\begin{bmatrix}6\\3\end{bmatrix},\quad t=3>0.
$$

The perspective image is
$$
P(z,t)=z/t.
$$

Compute each coordinate:
$$
\frac{z}{t}=\frac13\begin{bmatrix}6\\3\end{bmatrix}=\begin{bmatrix}2\\1\end{bmatrix}.
$$

The positive scale keeps the transformation in its domain.


```python
import sympy as sp

point = sp.Matrix([6, 3])
scale = sp.Integer(3)
image = point / scale

image
```
