### 1.1.7.2. Eigenvector


$$
(A - \lambda I) \mathbf{e}_{\lambda} = \mathbf{0}
$$


**Explanation:**

An eigenvector is a nonzero vector whose direction is unchanged by the matrix action. Algebraically, eigenvectors are the nonzero solutions of the null-space problem associated with an [eigenvalue](./01_eigenvalue.ipynb). The full collection for one eigenvalue is described in [eigenspaces](./03_eigenspace.ipynb).

**Example:**

For the eigenvalue $5$ of

$$
A =
\begin{bmatrix}
9 & -2 \\
-2 & 6
\end{bmatrix},
$$

one eigenvector is

$$
\mathbf{e}_5 =
\begin{bmatrix}
1 \\
2
\end{bmatrix}.
$$


```python
import sympy as sp

matrix_a = sp.Matrix([[9, -2], [-2, 6]])
eigenvector_data = matrix_a.eigenvects()

print("eigenvector_data =", eigenvector_data)
```
