### 1.1.5.1. Line in R^3


$$
\ell = \left\{ \mathbf{p}_0 + t \mathbf{v} \mid t \in \mathbb{R} \right\}
$$

$$
\frac{x - p_{0x}}{v_x} = \frac{y - p_{0y}}{v_y} = \frac{z - p_{0z}}{v_z}
$$


**Explanation:**

A line in $\mathbb{R}^3$ is the one-dimensional set traced by starting from a point $\mathbf{p}_0$ and moving along a direction vector $\mathbf{v}$. The parametric and symmetric equations are two equivalent ways to describe the same geometric object. This point-direction model is the geometric starting point for [planes](./02_plane_in_r3.ipynb), [projections](./07_projection_onto_a_line.ipynb), and [projection matrices](../06_Linear_Transformations/09_projection_matrix_representation.ipynb).

**Properties:**
- Scaling $\mathbf{v}$ by a nonzero constant does not change the line.
- Any two distinct points on the line determine a valid direction vector.

**Example:**

If

$$
\mathbf{p}_0 =
\begin{bmatrix}
1 \\
1 \\
1
\end{bmatrix},
\qquad
\mathbf{v} =
\begin{bmatrix}
1 \\
2 \\
3
\end{bmatrix},
$$

then

$$
\ell = \left\{
\begin{bmatrix}
1 \\
1 \\
1
\end{bmatrix}
+ t
\begin{bmatrix}
1 \\
2 \\
3
\end{bmatrix}
\mid t \in \mathbb{R}
\right\}
$$

and the symmetric equation is

$$
\frac{x - 1}{1} = \frac{y - 1}{2} = \frac{z - 1}{3}.
$$


```python
import sympy as sp

point_on_line = sp.Matrix([1, 1, 1])
line_direction = sp.Matrix([1, 2, 3])
parameter_values = [-1, 0, 2]
sample_points = [point_on_line + parameter_value * line_direction for parameter_value in parameter_values]

print("point_on_line =", list(point_on_line))
print("line_direction =", list(line_direction))
print("sample_points =", [list(sample_point) for sample_point in sample_points])
```
