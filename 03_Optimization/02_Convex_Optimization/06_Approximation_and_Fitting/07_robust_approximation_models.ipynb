### 3.2.6.7. Robust Approximation Models


$$\min_x\sup_{u\in\mathcal U}\|A(u)x-b\|$$


**Explanation:**

Robust approximation evaluates residuals against an uncertainty set rather than only at nominal data. The objective often becomes a worst-case [norm](../../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb), which remains convex for common norm-bounded uncertainty descriptions. Feasibility includes the uncertainty model; optimality controls the largest allowed residual.

The set $\mathcal U$ contains possible uncertainty values. For each decision $x$, the supremum takes the largest residual norm over that uncertainty set, so the minimization controls the worst allowed residual.

**Properties:**
- Worst-case objectives use a supremum over uncertainty.
- Robust models trade nominal fit for reliability.
- Norm-bounded uncertainty often produces tractable convex constraints.

**Intuition:**

The figure above motivates robust residual control; the calculation computes a worst-case scalar residual.

<img src="../../../Figures/03020607_bv_fig_6_15_robust_residuals.jpeg" alt="Boyd and Vandenberghe Figure 6.15: residuals under uncertainty for robust approximation" style="width: 60%; height: auto;">


**Numerical Example:**

The figure above motivates controlling large residuals. Let the uncertain residual be
$$
r(u)=2+u,
\qquad u\in[-1,1].
$$

At the endpoints,
$$
|r(-1)|=|2-1|=1,
$$
and
$$
|r(1)|=|2+1|=3.
$$

Therefore
$$
\sup_{u\in[-1,1]} |2+u|=3.
$$

The robust objective uses this worst-case value.


```python
import sympy as sp

nominal_residual = sp.Integer(2)
uncertainty_values = [sp.Integer(-1), sp.Integer(1)]
absolute_residuals = [abs(nominal_residual + uncertainty) for uncertainty in uncertainty_values]
worst_case_residual = max(absolute_residuals)

print("absolute_residuals =", absolute_residuals)
print("worst_case_residual =", worst_case_residual)
```
