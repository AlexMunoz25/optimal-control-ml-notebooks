### 3.1.9. Stepsize Rules


$$
\begin{aligned}
\alpha^k &\in \arg\min_{\alpha \ge 0} f(x^k+\alpha d^k), \\
f(x^k)-f(x^k+\beta^m s d^k)
&\ge -\sigma \beta^m s\nabla f(x^k)^{\mathsf T}d^k, \\
\sigma
&\le
\frac{f(x^k+\alpha^k d^k)-f(x^k)}{\alpha^k\nabla f(x^k)^{\mathsf T}d^k}
\le 1-\sigma.
\end{aligned}
$$


**Explanation:**

Once a descent direction has been chosen, the stepsize determines whether the algorithm makes useful progress. Exact line minimization is conceptually attractive because it chooses the best point along the ray, but it can be too expensive or unnecessary in practice. Sufficient-decrease rules therefore require enough decrease rather than exact minimization, especially the Armijo rule and the Goldstein rule. The common thread is the directional derivative $\nabla f(x^k)^{\mathsf T}d^k$, written with [matrix transpose](../../01_Foundations/01_Linear_Algebra/03_Matrix/06_matrix_transpose.ipynb) notation, which predicts the first-order decrease available along the chosen direction.

The first line defines exact line search: choose $\alpha$ to minimize the cost along the ray $x^k+\alpha d^k$. The later inequalities describe inexact stepsize tests that compare actual decrease with the directional derivative $\nabla f(x^k)^{\mathsf T}d^k$.

A beginner-friendly way to view these rules is as safeguards against two bad extremes. A step that is too large may overshoot and increase the cost; a step that is technically decreasing but too small may waste iterations and fail to drive the gradient toward zero. Sufficient decrease rules give a practical compromise between reliability and computational cost.

**Intuition:**

The first figure shows why accepting any cost-reducing step can still fail. Armijo backtracking fixes this by requiring the observed decrease to be large enough relative to the predicted linear decrease. Goldstein's rule places the step in an acceptable interval that avoids both overly short and overly long steps. The code cell implements the Armijo idea for a simple quadratic direction.

**Numerical Example:**

The Armijo figure shows that a step must reduce the cost enough, not merely be chosen along a descent direction. Let
$$
Q=\begin{bmatrix}5&0\\0&1\end{bmatrix},\qquad x^k=\begin{bmatrix}1\\2\end{bmatrix},\qquad d^k=-Qx^k=\begin{bmatrix}-5\\-2\end{bmatrix}.
$$

The current cost and directional derivative are
$$
f(x^k)=\frac{1}{2}(5\cdot 1^2+2^2)=4.5,
\qquad
\nabla f(x^k)^{\mathsf T}d^k=\begin{bmatrix}5&2\end{bmatrix}\begin{bmatrix}-5\\-2\end{bmatrix}=-29.
$$

With $\sigma=10^{-4}$, $\alpha=1$ gives
$$
x^k+d^k=\begin{bmatrix}-4\\0\end{bmatrix},
\qquad f(x^k+d^k)=\frac{1}{2}(5\cdot 16)=40.
$$

The observed decrease is $4.5-40=-35.5$, which is less than $-\sigma\alpha\nabla f(x^k)^{\mathsf T}d^k=0.0029$, so Armijo rejects the step. After two halvings, $\alpha=0.25$ gives
$$
x^k+0.25d^k=\begin{bmatrix}-0.25\\1.5\end{bmatrix},
\qquad f(x^k+0.25d^k)=\frac{1}{2}(5\cdot 0.25^2+1.5^2)=1.28125.
$$

Now $4.5-1.28125=3.21875 \ge 0.000725$, so the step is accepted.


```python
import casadi as ca

decision_variable = ca.SX.sym("x", 2)
Q = ca.DM([[5, 0], [0, 1]])
objective = 0.5 * ca.bilin(Q, decision_variable, decision_variable)
objective_function = ca.Function("f", [decision_variable], [objective])
gradient_function = ca.Function("grad", [decision_variable], [ca.gradient(objective, decision_variable)])

current_point = ca.DM([1, 2])
gradient = gradient_function(current_point)
direction = -gradient
trial_stepsize = 1.0
reduction_factor = 0.5
sufficient_decrease = 1e-4
directional_derivative = float(ca.dot(gradient, direction))
current_value = float(objective_function(current_point))

while current_value - float(objective_function(current_point + trial_stepsize * direction)) < -sufficient_decrease * trial_stepsize * directional_derivative:
    trial_stepsize *= reduction_factor

print("trial_stepsize =", trial_stepsize)
print("objective at accepted step =", objective_function(current_point + trial_stepsize * direction))
```
