### 1.1.7.7. Eigenvalues and Invertibility


$$
0 \notin \operatorname{eig}(A)
\Longleftrightarrow
A \text{ is invertible}
$$


**Explanation:**

Savov connects eigenvalues to invertibility through the determinant product formula. This notebook isolates the specific criterion that zero must not be an eigenvalue, instead of restating the broader [invertible matrix theorem](../06_Linear_Transformations/14_invertible_matrix_theorem.ipynb). The idea depends on the earlier notebook on [eigenvalues](./01_eigenvalue.ipynb).

**Example:**

If a matrix has eigenvalues $5$ and $10$, then it is invertible because neither eigenvalue is zero. If one eigenvalue were $0$, the matrix would have a nontrivial kernel and could not be inverted.


```python
import sympy as sp

matrix_a = sp.Matrix([[9, -2], [-2, 6]])
eigenvalues = list(matrix_a.eigenvals().keys())
is_invertible = 0 not in eigenvalues

print("eigenvalues =", eigenvalues)
print("is_invertible =", is_invertible)
```
