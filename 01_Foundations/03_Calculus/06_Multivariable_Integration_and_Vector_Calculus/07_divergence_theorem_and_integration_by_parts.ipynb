### 1.3.6.7. Divergence Theorem and Integration by Parts

$$
\iint_{S} \mathbf{F}\cdot\mathbf{N}\,dS = \iiint_{E} \nabla\cdot\mathbf{F}\,dV ;
$$

$$
\iiint_{E} u\,\Delta v\,dV = \iint_{S} u\,(\nabla v\cdot\mathbf{N})\,dS - \iiint_{E} \nabla u\cdot\nabla v\,dV
\quad (\text{Green's first identity}) .
$$

**Explanation:**

The divergence theorem equates the outward flux of a [vector field](./04_vector_fields_and_gradient_fields.ipynb) through a closed surface $S$ to the [triple integral](./02_triple_integrals.ipynb) of its [divergence](./05_divergence.ipynb) over the enclosed solid $E$ — the higher-dimensional [fundamental theorem of calculus](../03_Single_Variable_Integration/03_fundamental_theorem_of_calculus.ipynb). Applying it to $\mathbf{F} = u\nabla v$ yields Green's first identity, the multidimensional [integration by parts](../03_Single_Variable_Integration/05_integration_by_parts.ipynb): it moves a derivative from one factor to the other plus a boundary term. That move is the engine of the adjoint method for PDE-constrained optimal control and of weak/variational formulations.

**Properties:**
- Flux form of the FTC: integrating $\nabla\cdot\mathbf{F}$ over a volume reduces to boundary data on $S$.
- Green's identity is the divergence theorem applied to $u\nabla v$, splitting off a boundary term.

**Intuition:**

The outward flux of $\mathbf{F}$ through the closed surface $S$ equals the integral of its divergence over the enclosed solid $E$.

**Numerical Example:**

Verify the divergence theorem for $\mathbf{F} = (x, y, z)$ on the unit cube $E = [0,1]^3$. The divergence is

$$
\nabla\cdot\mathbf{F} = 1 + 1 + 1 = 3, \qquad
\iiint_{E} 3\,dV = 3\cdot\text{vol}(E) = 3\cdot 1 = 3 .
$$

The flux side: each of the six faces contributes $\int\int \mathbf{F}\cdot\mathbf{N}\,dS$. On the face $x = 1$ (outward normal $+\mathbf{i}$), $\mathbf{F}\cdot\mathbf{N} = x = 1$ over a unit-area face $\Rightarrow 1$; the face $x = 0$ gives $0$. The $y$- and $z$-pairs behave identically, so the total flux is $3\cdot(1) = 3$, matching the volume integral.

```python
import sympy as sp

x, y, z = sp.symbols("x y z")
field = (x, y, z)

divergence = sum(sp.diff(component, variable) for component, variable in zip(field, (x, y, z)))
volume_integral = sp.integrate(sp.integrate(sp.integrate(divergence, (z, 0, 1)), (y, 0, 1)), (x, 0, 1))

face_fluxes = {
    "x=1": sp.integrate(sp.integrate(field[0].subs(x, 1), (y, 0, 1)), (z, 0, 1)),
    "x=0": sp.integrate(sp.integrate(-field[0].subs(x, 0), (y, 0, 1)), (z, 0, 1)),
    "y=1": sp.integrate(sp.integrate(field[1].subs(y, 1), (x, 0, 1)), (z, 0, 1)),
    "y=0": sp.integrate(sp.integrate(-field[1].subs(y, 0), (x, 0, 1)), (z, 0, 1)),
    "z=1": sp.integrate(sp.integrate(field[2].subs(z, 1), (x, 0, 1)), (y, 0, 1)),
    "z=0": sp.integrate(sp.integrate(-field[2].subs(z, 0), (x, 0, 1)), (y, 0, 1)),
}
total_flux = sum(face_fluxes.values())

print("div F =", divergence)
print("∭_E div F dV =", volume_integral)
print("face fluxes  =", face_fluxes)
print("total surface flux =", total_flux, " (matches volume integral)")
```
