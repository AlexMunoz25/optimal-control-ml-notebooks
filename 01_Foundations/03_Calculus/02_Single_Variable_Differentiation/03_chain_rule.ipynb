### 1.3.2.3. Chain Rule

$$
h(x) = f\big(g(x)\big) \;\Longrightarrow\; h'(x) = f'\big(g(x)\big)\, g'(x)
\qquad\Big(\text{Leibniz: } \frac{dh}{dx} = \frac{dh}{du}\frac{du}{dx},\ u = g(x)\Big)
$$

**Explanation:**

The chain rule differentiates a composition: the derivative of the outer function evaluated at the inner function, times the derivative of the inner function. In Leibniz notation the intermediate variable's differentials "cancel," $\frac{dh}{dx} = \frac{dh}{du}\frac{du}{dx}$, which makes the rule easy to extend to long chains. Applied repeatedly through a chain of compositions it becomes the engine of [backpropagation](../07_Matrix_Calculus_and_Automatic_Differentiation/08_reverse_mode_automatic_differentiation.ipynb) and the [multivariable chain rule](../05_Multivariable_Differential_Calculus/08_multivariable_chain_rule.ipynb) used to train models.

**Properties:**
- Power rule for compositions: if $h(x) = \big(g(x)\big)^n$ then $h'(x) = n\big(g(x)\big)^{n-1} g'(x)$.
- For a chain $f_3 \circ f_2 \circ f_1$, the derivative is the product $f_3'\cdot f_2'\cdot f_1'$ evaluated along the chain.

**Numerical Example:**

Differentiate $h(x) = \sin(x^3)$. Take the outer $f(u) = \sin u$ and inner $u = g(x) = x^3$, so $f'(u) = \cos u$ and $g'(x) = 3x^2$:

$$
h'(x) = f'\big(g(x)\big)\, g'(x) = \cos(x^3)\cdot 3x^2 = 3x^2 \cos(x^3).
$$

For a power composition $p(x) = (2x^2 + 3)^5$, with $g(x) = 2x^2 + 3$ and $g'(x) = 4x$,

$$
p'(x) = 5\,(2x^2 + 3)^4 \cdot 4x = 20x\,(2x^2 + 3)^4 .
$$

```python
import sympy as sp

x, u = sp.symbols("x u")
outer = sp.sin(u)
inner = x**3

chain_rule_result = sp.diff(outer, u).subs(u, inner) * sp.diff(inner, x)
direct_result = sp.diff(sp.sin(x**3), x)

power_composition = (2 * x**2 + 3)**5
power_chain = sp.diff(power_composition, x)

print("f'(g(x)) * g'(x) =", chain_rule_result)
print("sympy d/dx sin(x^3) =", direct_result)
print("d/dx (2x^2+3)^5 =", sp.factor(power_chain))
```

**Equivalent CasADi implementation:**

A composition built as a CasADi graph is differentiated by reverse-mode accumulation through the chain, reproducing $3x^2\cos(x^3)$.

```python
import casadi as ca

x = ca.SX.sym("x")
composition = ca.sin(x**3)
chain_derivative = ca.gradient(composition, x)
derivative_function = ca.Function("dh", [x], [chain_derivative])

print("d/dx sin(x^3) =", chain_derivative)
print("value at x=1  :", float(derivative_function(1.0)))
```
