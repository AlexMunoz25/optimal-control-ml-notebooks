### 1.1.3.19. Matrix Block Multiplication

$$
\mathbf{A}\mathbf{B} =
\begin{bmatrix}
\mathbf{A}_{11}\mathbf{B}_{11} + \mathbf{A}_{12}\mathbf{B}_{21} &
\mathbf{A}_{11}\mathbf{B}_{12} + \mathbf{A}_{12}\mathbf{B}_{22} \\
\mathbf{A}_{21}\mathbf{B}_{11} + \mathbf{A}_{22}\mathbf{B}_{21} &
\mathbf{A}_{21}\mathbf{B}_{12} + \mathbf{A}_{22}\mathbf{B}_{22}
\end{bmatrix}
$$

**Explanation:**

Block multiplication generalizes matrix multiplication by partitioning matrices into submatrices.
This simplifies analysis of large matrices, enables parallel computation, and is useful in numerical linear algebra and control systems.

By grouping entries into submatrices, block multiplication applies the same rule as ordinary [matrix multiplication](./05_matrix_multiplication.ipynb) at a larger scale. It pairs naturally with the [matrix Kronecker product](./20_matrix_kronecker_product.ipynb), which also assembles large matrices from smaller pieces.

**Example:**

Partition

$$
\mathbf{A} =
\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 9
\end{bmatrix},
\quad
\mathbf{B} =
\begin{bmatrix}
9 & 8 & 7 \\
6 & 5 & 4 \\
3 & 2 & 1
\end{bmatrix}
$$

into $2 \times 2$ block matrices by rows and columns.
Multiplying the blocks yields the same result as standard multiplication.


```python
import sympy as sp

block_a11 = sp.Matrix([[1, 2], [3, 4]])
block_a12 = sp.Matrix([[5, 6], [7, 8]])
block_a21 = sp.Matrix([[9, 10], [11, 12]])
block_a22 = sp.Matrix([[13, 14], [15, 16]])

block_b11 = sp.Matrix([[17, 18], [19, 20]])
block_b12 = sp.Matrix([[21, 22], [23, 24]])
block_b21 = sp.Matrix([[25, 26], [27, 28]])
block_b22 = sp.Matrix([[29, 30], [31, 32]])

top_left = block_a11 * block_b11 + block_a12 * block_b21
top_right = block_a11 * block_b12 + block_a12 * block_b22
bottom_left = block_a21 * block_b11 + block_a22 * block_b21
bottom_right = block_a21 * block_b12 + block_a22 * block_b22

result = top_left.row_join(top_right).col_join(bottom_left.row_join(bottom_right))

print("Block multiplication result:")
sp.pprint(result)
```
