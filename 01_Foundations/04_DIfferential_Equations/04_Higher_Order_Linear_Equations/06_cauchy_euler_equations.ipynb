### 1.4.4.6. Cauchy-Euler Equations

$$
ax^2 y'' + bx y' + cy = 0 \;\xrightarrow{\,y = x^m\,}\; am(m-1) + bm + c = 0
\;\Leftrightarrow\; am^2 + (b - a)m + c = 0 .
$$

**Explanation:**

A **CauchyãEuler equation** has variable coefficients in which the power of $x$ matches the order of the derivative, $a_k x^k y^{(k)}$. Trying $y = x^m$ reduces it to an **auxiliary equation** in $m$, exactly as $e^{mx}$ does for [constant coefficients](./03_homogeneous_equations_with_constant_coefficients.ipynb). The three root cases mirror the constant-coefficient ones: distinct real roots give $x^{m_1}, x^{m_2}$; a repeated root gives $x^{m_1}, x^{m_1}\ln x$; complex roots $\alpha \pm i\beta$ give $x^\alpha\cos(\beta\ln x)$ and $x^\alpha\sin(\beta\ln x)$. Solutions are studied on $(0, \infty)$ since the lead coefficient vanishes at $x = 0$.

**Properties:**
- The substitution $x = e^t$ transforms a CauchyãEuler equation into a constant-coefficient one in $t$.
- Repeated and complex cases parallel the constant-coefficient results with $x^m$, $\ln x$ replacing $e^{mx}$, $x$.

**Numerical Example:**

Solve $x^2 y'' - 2x y' - 4y = 0$ on $(0, \infty)$.

**Substitute** $y = x^m$, so $y' = m x^{m-1}$ and $y'' = m(m-1)x^{m-2}$:

$$
x^2\,m(m-1)x^{m-2} - 2x\,m x^{m-1} - 4x^m = x^m\big(m(m-1) - 2m - 4\big) = x^m(m^2 - 3m - 4).
$$

**Auxiliary equation.** Since $x^m \neq 0$, set

$$
m^2 - 3m - 4 = (m + 1)(m - 4) = 0 \;\Rightarrow\; m_1 = -1,\ m_2 = 4 .
$$

**General solution** (distinct real roots):

$$
y = c_1 x^{-1} + c_2 x^{4} .
$$

```python
import sympy as sp

x, m = sp.symbols("x m")
y = sp.Function("y")

substitution = sp.expand((x**2 * sp.diff(x**m, x, 2) - 2 * x * sp.diff(x**m, x) - 4 * x**m) / x**m)
auxiliary_roots = sp.roots(substitution, m)
general = sp.dsolve(sp.Eq(x**2 * y(x).diff(x, 2) - 2 * x * y(x).diff(x) - 4 * y(x), 0), y(x))

print("after y = x^m, bracket =", substitution, "= 0")
print("auxiliary roots m =", auxiliary_roots)
print("general solution:", general.rhs)
```
