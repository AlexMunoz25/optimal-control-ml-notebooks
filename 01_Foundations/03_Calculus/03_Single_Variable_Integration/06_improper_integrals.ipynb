### 1.3.3.6. Improper Integrals

$$
\int_{a}^{\infty} f(x)\,dx = \lim_{t \to \infty} \int_{a}^{t} f(x)\,dx,
\qquad
\int_{a}^{b} f(x)\,dx = \lim_{t \to b^{-}} \int_{a}^{t} f(x)\,dx \ \ (\text{singularity at } b).
$$

**Explanation:**

An improper integral extends the definite integral to an unbounded interval or an unbounded integrand by taking a [limit](../01_Limits_and_Continuity/03_one_sided_limits_and_limits_at_infinity.ipynb) of ordinary integrals. If the limit is a finite number the integral *converges*; otherwise it *diverges*. These integrals are unavoidable in probability and machine learning, where densities are integrated over all of $\mathbb{R}$: a normalizing constant, an expectation $\int x\,p(x)\,dx$, or the total probability $\int p(x)\,dx = 1$ are all improper integrals.

**Properties:**
- $\displaystyle\int_{1}^{\infty} \frac{1}{x^{p}}\,dx$ converges iff $p > 1$; the borderline $p = 1$ diverges.
- The Gaussian integral $\displaystyle\int_{-\infty}^{\infty} e^{-x^{2}}\,dx = \sqrt{\pi}$ underlies the normalization of the normal distribution.

**Intuition:**

The integral over an unbounded interval is the limit of ordinary integrals up to $t$ as the upper endpoint $t$ moves outward.

**Numerical Example:**

Evaluate $\int_{1}^{\infty} \dfrac{1}{x^{2}}\,dx$ as a limit. For finite $t$,

$$
\int_{1}^{t} x^{-2}\,dx = \left[-\frac{1}{x}\right]_{1}^{t} = 1 - \frac{1}{t} .
$$

Letting $t \to \infty$, the term $1/t \to 0$, so the integral **converges** to

$$
\int_{1}^{\infty} \frac{1}{x^2}\,dx = \lim_{t\to\infty}\left(1 - \frac{1}{t}\right) = 1 .
$$

By contrast $\int_{1}^{\infty} \tfrac1x\,dx = \lim_{t\to\infty}\ln t = \infty$ diverges, illustrating the $p > 1$ threshold.

```python
import sympy as sp

x, t = sp.symbols("x t", positive=True)

finite_integral = sp.integrate(1 / x**2, (x, 1, t))
convergent_value = sp.limit(finite_integral, t, sp.oo)
divergent_value = sp.integrate(1 / x, (x, 1, sp.oo))
gaussian_integral = sp.integrate(sp.exp(-x**2), (x, -sp.oo, sp.oo))

print("∫_1^t x^-2 dx =", finite_integral)
print("∫_1^∞ x^-2 dx =", convergent_value, " (converges)")
print("∫_1^∞ x^-1 dx =", divergent_value, " (diverges)")
print("∫_-∞^∞ e^{-x^2} dx =", gaussian_integral)
```
