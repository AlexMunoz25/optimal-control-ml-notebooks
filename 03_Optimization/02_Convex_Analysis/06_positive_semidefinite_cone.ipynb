### 3.2.6. Positive Semidefinite Cone


$$\mathbf{S}_+^n=\{X\in\mathbf{S}^n\mid z^TXz\geq0\ \forall z\}$$


**Explanation:**

The [positive semidefinite cone](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/18_matrix_definiteness.ipynb) consists of symmetric matrices whose quadratic forms are nonnegative in every direction. In small examples, [eigenvalues](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/08_eigenvalue.ipynb) give a direct membership test because a symmetric matrix is positive semidefinite exactly when all eigenvalues are nonnegative.

The symbol $X$ is a symmetric [matrix](../../01_Foundations/01_Linear_Algebra/03_Matrix/01_matrix_definition.ipynb) and $z$ ranges over vectors. The inequality $z^TXz\geq0$ says every quadratic form generated by $X$ is nonnegative, which is the positive semidefinite condition in [matrix definiteness](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/18_matrix_definiteness.ipynb).

**Properties:**
- All eigenvalues must be nonnegative for positive semidefiniteness.
- Positive semidefinite matrices form a convex [cone](03_cones_and_conic_combinations.ipynb).
- Matrix inequalities use this cone as an order relation.

**Intuition:**

The figure above gives cone geometry; the arithmetic checks membership for one concrete matrix.

**Numerical Example:**

The figure above gives cone geometry; the arithmetic checks membership for one concrete matrix.

Let
$$
X=\begin{bmatrix}2&1\\1&2\end{bmatrix}.
$$

The characteristic polynomial is
$$
\det(X-\lambda I)=(2-\lambda)^2-1.
$$

Expanding gives
$$
(2-\lambda)^2-1=\lambda^2-4\lambda+3=(\lambda-1)(\lambda-3).
$$

Thus the eigenvalues are $1$ and $3$. Since both are nonnegative, $X\succeq0$.


```python
import sympy as sp

matrix = sp.Matrix([[2, 1], [1, 2]])
eigenvalues = matrix.eigenvals()

print("matrix =", matrix)
print("eigenvalues =", eigenvalues)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

matrix = ca.DM([[2, 1], [1, 2]])
eigenvalues = ca.eig_symbolic(ca.SX(matrix))

print("matrix =", matrix)
print("eigenvalues =", eigenvalues)
```
