### 3.2.8.4. Distance, Angle, and Gram Matrix Problems


$$G_{ij}=x_i^Tx_j$$


**Explanation:**

A Gram matrix stores pairwise inner products of vectors. Distances, angles, and realizability constraints can be expressed through this matrix, with [positive semidefiniteness](../../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/18_matrix_definiteness.ipynb) as the key convex condition. Feasibility is therefore a matrix [cone](../02_Convex_Sets/03_cones_and_conic_combinations.ipynb) constraint plus any specified geometric entries.

Each Gram matrix entry $G_{ij}$ is an [inner product](../../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/02_inner_product_space.ipynb) between vectors $x_i$ and $x_j$. The full matrix records lengths on the diagonal and pairwise angle information off the diagonal.

**Properties:**
- Diagonal entries are squared [norms](../../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb).
- Off-diagonal entries are inner products.
- A valid Gram matrix must be positive semidefinite.

**Intuition:**

The example builds the Gram matrix from two simple vectors.


**Numerical Example:**

Let
$$
v_1=\begin{bmatrix}1\\0\end{bmatrix},\quad
v_2=\begin{bmatrix}0\\2\end{bmatrix}.
$$

The Gram entries are
$$
G_{11}=v_1^Tv_1=1,
\qquad
G_{12}=v_1^Tv_2=0,
\qquad
G_{22}=v_2^Tv_2=4.
$$

Therefore
$$
G=\begin{bmatrix}1&0\\0&4\end{bmatrix}.
$$


```python
import sympy as sp

vector_1 = sp.Matrix([1, 0])
vector_2 = sp.Matrix([0, 2])
gram = sp.Matrix([[(vector_1.T * vector_1)[0], (vector_1.T * vector_2)[0]], [(vector_2.T * vector_1)[0], (vector_2.T * vector_2)[0]]])

gram
```
