### 1.4.7.2. Method of Frobenius

$$
y = \sum_{n=0}^{\infty} c_n x^{n+r},
\qquad
\text{indicial equation in } r \ \Rightarrow\ \text{recurrence for } c_n .
$$

**Explanation:**

At a **regular singular point** the ordinary [power-series method](./01_power_series_solutions_about_ordinary_points.ipynb) can fail, but **Frobenius' theorem** guarantees at least one solution of the generalized form $y = \sum c_n x^{n+r}$ with an extra exponent $r$ to be found. Substituting and collecting the lowest power of $x$ yields the **indicial equation** for $r$; its roots set the leading behaviour, and a recurrence then generates the coefficients (a different recurrence for each root). When the roots differ by a non-integer, both lead to independent Frobenius series. This is how the special functions of mathematical physics (Bessel, etc.) arise from singular models.

**Properties:**
- $x_0$ is a *regular* singular point if $(x-x_0)P(x)$ and $(x-x_0)^2 Q(x)$ are analytic there.
- The indicial equation is the coefficient of the lowest power of $x$ set to zero; its roots are the admissible exponents $r$.

**Numerical Example:**

Solve $3xy'' + y' - y = 0$ ($x = 0$ is a regular singular point).

**Substitute** $y = \sum_{n=0}^{\infty} c_n x^{n+r}$. Then $3xy'' + y' - y$ becomes

$$
x^r\!\left[r(3r-2)c_0\,x^{-1} + \sum_{k=0}^{\infty}\big[(k+r+1)(3k+3r+1)c_{k+1} - c_k\big]x^k\right] = 0 .
$$

**Indicial equation** (coefficient of $x^{r-1}$, with $c_0 \neq 0$):

$$
r(3r - 2) = 0 \;\Rightarrow\; r_1 = \tfrac{2}{3},\quad r_2 = 0 .
$$

**Recurrence:** $\displaystyle c_{k+1} = \frac{c_k}{(k+r+1)(3k+3r+1)}$.

For $r_1 = \tfrac23$: $c_1 = \tfrac{c_0}{5}$, $c_2 = \tfrac{c_0}{80}$, …, giving $c_n = \dfrac{c_0}{n!\,5\cdot 8\cdots(3n+2)}$, so

$$
y_1 = x^{2/3}\!\left(1 + \tfrac15 x + \tfrac{1}{80}x^2 + \cdots\right).
$$

For $r_2 = 0$: $c_1 = c_0$, $c_2 = \tfrac{c_0}{8}$, …, giving $c_n = \dfrac{c_0}{n!\,1\cdot 4\cdots(3n-2)}$, so

$$
y_2 = 1 + x + \tfrac18 x^2 + \cdots .
$$

Since $r_1 - r_2 = \tfrac23$ is not an integer, $y = C_1 y_1 + C_2 y_2$ is the general solution.

```python
import sympy as sp

x, r = sp.symbols("x r")

indicial_roots = sp.solve(r * (3 * r - 2), r)

def frobenius_coefficients(exponent, count=4):
    values = [sp.Integer(1)]
    for k in range(count - 1):
        values.append(sp.nsimplify(values[k] / ((k + exponent + 1) * (3 * k + 3 * exponent + 1))))
    return values

print("indicial equation r(3r - 2) = 0  ->  r =", indicial_roots)
for exponent in indicial_roots:
    coefficients = frobenius_coefficients(exponent)
    print(f"r = {exponent}: c0..c3 =", coefficients)
```
