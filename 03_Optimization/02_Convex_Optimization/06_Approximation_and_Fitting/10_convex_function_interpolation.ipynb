### 3.2.6.10. Convex Function Interpolation


$$f(x_j)\geq f(x_i)+g_i^T(x_j-x_i)$$


**Explanation:**

[Convex function](../03_Convex_Functions/01_convex_function_definition_and_examples.ipynb) interpolation asks whether finite data can come from a convex function. First-order convexity inequalities between sampled points provide the finite consistency conditions used to check the data.

The points $x_i$ and $x_j$ are sample locations, and $g_i$ is a subgradient assigned at $x_i$. The inequality says every other sample value must lie above the affine support plane built from $x_i$, $f(x_i)$, and $g_i$.

**Properties:**
- The gradient defines the tangent approximation.
- Convexity keeps the true function above that approximation.
- At an optimum, the first-order inequality holds against every feasible point.

**Intuition:**

The computation compares a tangent lower bound with the true value.


**Numerical Example:**

Let $f(x)=x^2$, expand around $x=1$, and compare with $y=3$. Since $f'(x)=2x$,
$$
f'(1)=2.
$$

The tangent lower bound at $y=3$ is
$$
f(1)+f'(1)(3-1)=1^2+2\cdot2=5.
$$

The true value is
$$
f(3)=3^2=9.
$$

Because $9\geq5$, the tangent line gives a valid global lower bound.


```python
import sympy as sp

point_x = sp.Integer(1)
point_y = sp.Integer(3)
lower_bound = point_x ** 2 + 2 * point_x * (point_y - point_x)
true_value = point_y ** 2

print("lower_bound =", lower_bound)
print("true_value =", true_value)
```
