### 1.1.6.11. Rotation Matrix Representation


$$
M_{R_{\theta}} =
\begin{bmatrix}
\cos \theta & -\sin \theta \\
\sin \theta & \cos \theta
\end{bmatrix}
$$


**Explanation:**

A planar rotation is the [linear transformation](./01_linear_transformation.ipynb) that turns every vector by the same angle while preserving length. Probing the standard basis produces its matrix representation, which also foreshadows the later viewpoint of an [orthogonal matrix](../07_Theoretical_Linear_Algebra/11_orthogonal_matrix.ipynb). The coordinate-side interpretation is related to [coordinate-system transformations](../08_Coordinate_Transformations/01_coordinate_system_transformation.ipynb).

**Example:**

For $\theta = \pi / 6$,

$$
M_{R_{\theta}} =
\begin{bmatrix}
\sqrt{3}/2 & -1/2 \\
1/2 & \sqrt{3}/2
\end{bmatrix}.
$$


```python
import sympy as sp

rotation_angle = sp.pi / 6
rotation_matrix = sp.Matrix(
    [
        [sp.cos(rotation_angle), -sp.sin(rotation_angle)],
        [sp.sin(rotation_angle), sp.cos(rotation_angle)],
    ]
)

print("rotation_matrix =")
sp.pprint(rotation_matrix)
```
