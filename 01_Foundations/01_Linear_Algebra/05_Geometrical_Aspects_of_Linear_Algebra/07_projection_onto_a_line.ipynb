### 1.1.5.7. Projection onto a Line


$$
\Pi_{\ell}(\mathbf{u}) =
\frac{\mathbf{u} \cdot \mathbf{v}}{\lVert \mathbf{v} \rVert^2} \mathbf{v}
$$


**Explanation:**

A projection onto a line extracts the part of a vector that points along the chosen direction. The projected vector is the along-line component, while the difference $\mathbf{u} - \Pi_{\ell}(\mathbf{u})$ is the perpendicular remainder. The formula depends on the [dot product](../02_Vector/04_vector_dot_product.ipynb), which converts directional information into a scalar coefficient.

<p align="center">
  <img src="../../../Figures/01010204_scalar_product_projection.jpeg" alt="Projection of one vector onto another" width="420">
</p>

**Example:**

Project

$$
\mathbf{u}=\begin{bmatrix}3\\4\end{bmatrix}
$$

onto the line direction

$$
\mathbf{d}=\begin{bmatrix}1\\0\end{bmatrix}.
$$

Then

$$
\operatorname{proj}_{\mathbf{d}}\mathbf{u}
=
\frac{\mathbf{u}\cdot\mathbf{d}}{\mathbf{d}\cdot\mathbf{d}}\mathbf{d}
=
\frac{3\cdot1+4\cdot0}{1^2+0^2}\begin{bmatrix}1\\0\end{bmatrix}
=3\begin{bmatrix}1\\0\end{bmatrix}
=\begin{bmatrix}3\\0\end{bmatrix}.
$$

The remainder is

$$
\mathbf{u}-\operatorname{proj}_{\mathbf{d}}\mathbf{u}
=\begin{bmatrix}3\\4\end{bmatrix}-\begin{bmatrix}3\\0\end{bmatrix}
=\begin{bmatrix}0\\4\end{bmatrix}.
$$


```python
import sympy as sp

vector_u = sp.Matrix([3, 4])
line_direction = sp.Matrix([1, 0])
projection_value = vector_u.dot(line_direction) / line_direction.dot(line_direction) * line_direction
remainder_value = vector_u - projection_value

print("projection_value =", projection_value)
print("remainder_value =", remainder_value)
```
