### 1.1.5.7. Projection onto a Line


$$
\Pi_{\ell}(\mathbf{u}) =
\frac{\mathbf{u} \cdot \mathbf{v}}{\lVert \mathbf{v} \rVert^2} \mathbf{v}
$$


**Explanation:**

A projection onto a line extracts the part of a vector that points along the chosen direction. Savov emphasizes the picture-proof interpretation: the projected vector is the along-line component, while the difference $\mathbf{u} - \Pi_{\ell}(\mathbf{u})$ is the perpendicular remainder. The concept depends on the [dot product](../02_Vector/04_vector_dot_product.ipynb), so this notebook links there instead of rederiving scalar projection.

**Example:**

If

$$
\mathbf{u} =
\begin{bmatrix}
3 \\
4
\end{bmatrix},
\qquad
\mathbf{v} =
\begin{bmatrix}
1 \\
0
\end{bmatrix},
$$

then

$$
\Pi_{\ell}(\mathbf{u}) =
\begin{bmatrix}
3 \\
0
\end{bmatrix}.
$$


```python
import sympy as sp

vector_u = sp.Matrix([3, 4])
line_direction = sp.Matrix([1, 0])
projection_value = vector_u.dot(line_direction) / line_direction.dot(line_direction) * line_direction
remainder_value = vector_u - projection_value

print("projection_value =", projection_value)
print("remainder_value =", remainder_value)
```
