### A2.3.1. New Pass Manager

> *The new pass manager (NPM) organizes optimization and analysis passes into a hierarchy of pass managers — module, CGSCC, function, and loop — each running passes at the appropriate granularity.*

**Explanation:**

LLVM's **New Pass Manager** (introduced as default in LLVM 14) replaces the legacy pass manager with a cleaner design based on explicit pass pipelines and analysis invalidation.

**Pass Manager Hierarchy:**

1. **ModulePassManager** — runs passes over the entire module.
2. **CGSCCPassManager** — runs passes over call-graph strongly connected components.
3. **FunctionPassManager** — runs passes over individual functions.
4. **LoopPassManager** — runs passes over individual loops.

**Key Concepts:**

- **AnalysisManager** — caches analysis results and tracks invalidation. A pass requests an analysis via `AM.getResult<AnalysisT>(F)`.
- **PreservedAnalyses** — returned by each pass to indicate which analyses remain valid after the transformation.
- **Pass Pipeline** — a textual description of the pass ordering: `opt -passes='function(instcombine,simplifycfg)'`.

**Example:**

```cpp
struct MyPass : public PassInfoMixin<MyPass> {
  PreservedAnalyses run(Function &F, FunctionAnalysisManager &AM) {
    // transformation logic
    return PreservedAnalyses::all();
  }
};
```

````python
pass_hierarchy = [
    ("ModulePassManager", "Entire module", "GlobalDCE, Inliner"),
    ("CGSCCPassManager", "Call-graph SCC", "Inliner, ArgumentPromotion"),
    ("FunctionPassManager", "Single function", "InstCombine, SimplifyCFG, GVN"),
    ("LoopPassManager", "Single loop", "LICM, LoopUnroll, IndVarSimplify"),
]

print("New Pass Manager Hierarchy:")
print(f"{'Manager':<24} {'Scope':<20} {'Example Passes'}")
print("-" * 72)
for manager, scope, examples in pass_hierarchy:
    print(f"{manager:<24} {scope:<20} {examples}")

pipeline_example = "module(cgscc(inline),function(instcombine,simplifycfg,loop(licm)))"
print(f"\nExample pipeline: opt -passes='{pipeline_example}'")
````
