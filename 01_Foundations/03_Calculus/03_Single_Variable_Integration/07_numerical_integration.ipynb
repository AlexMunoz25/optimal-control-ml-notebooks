### 1.3.3.7. Numerical Integration

$$
\textbf{Trapezoidal:}\quad \int_{a}^{b} f\,dx \approx \frac{\Delta x}{2}\Big(f_0 + 2f_1 + \cdots + 2f_{n-1} + f_n\Big),
$$

$$
\textbf{Simpson:}\quad \int_{a}^{b} f\,dx \approx \frac{\Delta x}{3}\Big(f_0 + 4f_1 + 2f_2 + 4f_3 + \cdots + 4f_{n-1} + f_n\Big),\ \ n \text{ even}.
$$

**Explanation:**

When an antiderivative is unavailable, definite integrals are approximated by sampling $f$ on a grid and combining the values with fixed weights. The trapezoidal rule joins consecutive samples with straight lines (piecewise-linear model); Simpson's rule fits a parabola through each pair of subintervals (piecewise-quadratic model) and is far more accurate for smooth integrands. These quadrature rules are the discretization behind [cost integrals](../../../04_Optimal_Control) in trajectory optimization, collocation methods, and Monte-Carlo-free estimates of returns.

**Properties:**
- Error scales as $O(\Delta x^{2})$ for the trapezoidal rule and $O(\Delta x^{4})$ for Simpson's rule.
- Simpson's rule integrates cubics exactly despite using only quadratic pieces.

**Intuition:**

The trapezoidal rule joins consecutive samples with straight lines, while Simpson's rule fits a parabola through each pair of subintervals.

**Numerical Example:**

Approximate $\int_{0}^{2} x^2\,dx$ (exact value $\tfrac83 \approx 2.6667$) with $n = 4$, $\Delta x = 0.5$, samples $f = 0, 0.25, 1, 2.25, 4$.

Trapezoidal:
$$
\frac{0.5}{2}\big(0 + 2(0.25) + 2(1) + 2(2.25) + 4\big) = 0.25\,(11) = 2.75 .
$$

Simpson:
$$
\frac{0.5}{3}\big(0 + 4(0.25) + 2(1) + 4(2.25) + 4\big) = \frac{0.5}{3}(16) = 2.6667 ,
$$

which is *exact* here because $x^2$ is a polynomial of degree $\le 3$.

```python
import sympy as sp

x = sp.symbols("x")
f = x**2
left_end, right_end, subinterval_count = 0, 2, 4
width = sp.Rational(right_end - left_end, subinterval_count)
samples = [f.subs(x, left_end + width * index) for index in range(subinterval_count + 1)]

trapezoidal = width / 2 * (samples[0] + 2 * sum(samples[1:-1]) + samples[-1])
simpson_weights = [1] + [4 if index % 2 else 2 for index in range(1, subinterval_count)] + [1]
simpson = width / 3 * sum(weight * value for weight, value in zip(simpson_weights, samples))
exact_integral = sp.integrate(f, (x, left_end, right_end))

print("trapezoidal =", trapezoidal, "=", float(trapezoidal))
print("simpson     =", simpson, "=", float(simpson))
print("exact       =", exact_integral, "=", float(exact_integral))
```

**Equivalent SciPy implementation:**

SciPy provides the same quadrature rules as array routines, matching the hand computation.

```python
import numpy as np
from scipy import integrate

grid = np.linspace(0, 2, 5)
values = grid**2

print("trapezoid (scipy) =", integrate.trapezoid(values, grid))
print("simpson   (scipy) =", integrate.simpson(values, x=grid))
```
