### 3.2.8.2. Separating Points from Convex Sets


$$a=P_C(x_0)-x_0$$


**Explanation:**

A point outside a closed [convex set](../02_Convex_Sets/01_affine_sets_and_convex_sets.ipynb) can be separated from the set by a hyperplane. The projection residual gives the separating normal in Euclidean geometry. Feasibility is membership in the convex set, and separation is the certificate that the outside point is not feasible.

The vector $a$ is the difference between the projection of $x_0$ on $C$ and the original point $x_0$. For Euclidean projection, this residual vector is normal to a separating hyperplane.

**Properties:**
- The separator places the convex set on one side.
- The outside point lies on the other side.
- Projection gives a constructive separator.

**Intuition:**

The example places a scalar threshold between an interval and an outside point.


**Numerical Example:**

Let
$$
C=[-1,1]
$$
and take the outside point $x_0=3$.

A separating threshold is
$$
b=2.
$$

Every point $x\in C$ satisfies
$$
x\leq1\leq2,
$$
while the outside point satisfies
$$
3\geq2.
$$

Therefore the scalar hyperplane $x=2$ separates $C$ from $x_0$.


```python
import sympy as sp

interval_upper_bound = sp.Integer(1)
outside_point = sp.Integer(3)
separator = sp.Integer(2)

print("interval_upper_bound <= separator =", interval_upper_bound <= separator)
print("separator <= outside_point =", separator <= outside_point)
```
