### 3.3.8. Semidefinite Programming and LMIs


$$F(x)=F_0+x_1F_1+\cdots+x_nF_n\succeq0$$


**Explanation:**

Semidefinite programs optimize a linear objective subject to affine matrix inequalities. Each linear matrix inequality requires an affine symmetric matrix to lie in the [positive semidefinite cone](../02_Convex_Analysis/06_positive_semidefinite_cone.ipynb).

The matrix $F(x)$ is affine in the scalar variables $x_i$. The condition $F(x)\succeq0$ means this symmetric matrix must be positive semidefinite, which is a [matrix definiteness](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/18_matrix_definiteness.ipynb) constraint.

**Properties:**
- All [eigenvalues](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/08_eigenvalue.ipynb) must be nonnegative for positive semidefiniteness.
- Positive semidefinite matrices form a convex [cone](../02_Convex_Analysis/03_cones_and_conic_combinations.ipynb).
- Matrix inequalities use this cone as an order relation.

**Intuition:**

The example checks a concrete two-by-two matrix through its eigenvalues.


**Numerical Example:**

Let
$$
X=\begin{bmatrix}2&1\\1&2\end{bmatrix}.
$$

The characteristic polynomial is
$$
\det(X-\lambda I)=(2-\lambda)^2-1.
$$

Expanding gives
$$
(2-\lambda)^2-1=\lambda^2-4\lambda+3=(\lambda-1)(\lambda-3).
$$

Thus the eigenvalues are $1$ and $3$. Since both are nonnegative, $X\succeq0$.


```python
import sympy as sp

matrix = sp.Matrix([[2, 1], [1, 2]])
eigenvalues = matrix.eigenvals()

print("matrix =", matrix)
print("eigenvalues =", eigenvalues)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

matrix = ca.DM([[2, 1], [1, 2]])
eigenvalues = ca.eig_symbolic(ca.SX(matrix))

print("matrix =", matrix)
print("eigenvalues =", eigenvalues)
```

**Feasible-set geometry:**

A semidefinite program constrains an affine symmetric matrix to the positive semidefinite cone. For the $2\times2$ matrices $\begin{bmatrix}x&y\\y&z\end{bmatrix}$ the boundary of this cone is the rank-one surface $xz=y^2$ with $x,z\ge0$, a curved cone with one flat face. Intersecting it with a polyhedron, an ellipsoid, and a second-order cone bounds the feasible set into a smooth curved body. The two cells below redraw the cone boundary and one such bounded intersection.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LightSource

boundary_scale = np.linspace(0, 1, 50)
boundary_angle = np.linspace(0, np.pi, 90)
scale_grid, angle_grid = np.meshgrid(boundary_scale, boundary_angle)
cone_x = scale_grid * np.cos(angle_grid) ** 2
cone_y = scale_grid * np.cos(angle_grid) * np.sin(angle_grid)
cone_z = scale_grid * np.sin(angle_grid) ** 2

cap_fraction = np.linspace(0, 1, 20)
rim_angle = np.linspace(0, np.pi, 90)
rim_grid, fraction_grid = np.meshgrid(rim_angle, cap_fraction)
rim_x = np.cos(rim_grid) ** 2
rim_y = np.cos(rim_grid) * np.sin(rim_grid)
rim_z = np.sin(rim_grid) ** 2
cap_center_x, cap_center_z = 0.5, 0.5
cap_x = cap_center_x + fraction_grid * (rim_x - cap_center_x)
cap_y = fraction_grid * rim_y
cap_z = cap_center_z + fraction_grid * (rim_z - cap_center_z)

light = LightSource(azdeg=60, altdeg=45)
figure = plt.figure(figsize=(5, 5))
axis = figure.add_subplot(111, projection="3d")
axis.plot_surface(cone_x, cone_y, cone_z, color="#E08E2F", lightsource=light, linewidth=0, antialiased=True, shade=True)
axis.plot_surface(cap_x, cap_y, cap_z, color="#e89a3c", linewidth=0, shade=False)
axis.set_xlim(0, 1)
axis.set_ylim(-0.5, 0.5)
axis.set_zlim(0, 1)
axis.view_init(elev=18, azim=-60)
for pane in (axis.xaxis.pane, axis.yaxis.pane, axis.zaxis.pane):
    pane.set_facecolor((1, 1, 1, 0))
    pane.set_edgecolor("#cfcfcf")
axis.set_xticklabels([])
axis.set_yticklabels([])
axis.set_zticklabels([])
axis.tick_params(length=0)
figure.savefig("../../Figures/030308_sdp_positive_semidefinite_cone.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LightSource

radius = np.linspace(0, 1, 60)
angle = np.linspace(0, 2 * np.pi, 90)
radius_grid, angle_grid = np.meshgrid(radius, angle)
semi_axis_x, semi_axis_y, semi_axis_z = 1.0, 0.85, 0.7
cap_x = semi_axis_x * radius_grid * np.cos(angle_grid)
cap_y = semi_axis_y * radius_grid * np.sin(angle_grid)
cap_z = semi_axis_z * np.sqrt(np.clip(1 - radius_grid ** 2, 0, 1))

light = LightSource(azdeg=45, altdeg=55)
figure = plt.figure(figsize=(5, 5))
axis = figure.add_subplot(111, projection="3d")
axis.plot_surface(cap_x, cap_y, cap_z, color="#E08E2F", lightsource=light, linewidth=0, antialiased=True, shade=True)
axis.set_xlim(-1, 1)
axis.set_ylim(-1, 1)
axis.set_zlim(0, 1)
axis.view_init(elev=12, azim=-75)
axis.set_box_aspect((1, 1, 0.55))
for pane in (axis.xaxis.pane, axis.yaxis.pane, axis.zaxis.pane):
    pane.set_facecolor((1, 1, 1, 0))
    pane.set_edgecolor("#cfcfcf")
axis.set_xticklabels([])
axis.set_yticklabels([])
axis.set_zticklabels([])
axis.tick_params(length=0)
figure.savefig("../../Figures/030308_sdp_combined_feasible_region.png", dpi=150, bbox_inches="tight")
plt.show()
```
