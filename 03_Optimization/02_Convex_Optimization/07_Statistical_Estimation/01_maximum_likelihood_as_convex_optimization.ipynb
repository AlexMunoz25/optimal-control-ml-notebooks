### 3.2.7.1. Maximum Likelihood as Convex Optimization


$$\max_\theta\sum_i\log p(y_i;\theta)$$


**Explanation:**

Maximum likelihood chooses parameters that make the observed data most probable. Many estimation problems become convex after minimizing the negative log-likelihood over a convex parameter domain. Feasibility is the parameter domain, and optimality is the point where the log-likelihood can no longer be increased.

The parameter $\theta$ is the decision variable, and $y_i$ are observed data. The objective sums log-likelihood terms, so maximizing it chooses the parameter value that makes the observed data most likely under the model.

**Properties:**
- Log-likelihoods turn products of probabilities into sums.
- The Bernoulli parameter must satisfy 0<p<1.
- For simple Bernoulli data, the optimizer is the observed success fraction.

**Intuition:**

The numerical example derives the estimate from first-order optimality instead of just quoting the sample proportion.


**Numerical Example:**

Observe four Bernoulli samples with three successes and one failure. If $p$ is the success probability, then
$$
\ell(p)=3\log p+\log(1-p),\qquad 0<p<1.
$$

Differentiate:
$$
\ell'(p)=\frac3p-\frac{1}{1-p}.
$$

Set $\ell'(p)=0$:
$$
\frac3p=\frac{1}{1-p}.
$$

Cross-multiply:
$$
3(1-p)=p.
$$

Thus
$$
3-3p=p,\qquad 4p=3,\qquad p^\star=\frac34.
$$

The maximum likelihood estimate is the observed success fraction.


```python
import sympy as sp

probability = sp.symbols('p')
log_likelihood = 3 * sp.log(probability) + sp.log(1 - probability)
stationary_point = sp.solve(sp.diff(log_likelihood, probability), probability)[0]

print("stationary_point =", stationary_point)
```

**Direct `casadi` implementation:**

```python
import casadi as ca

opti = ca.Opti()
probability = opti.variable()
opti.minimize(-(3 * ca.log(probability) + ca.log(1 - probability)))
opti.subject_to(probability > 0)
opti.subject_to(probability < 1)
opti.set_initial(probability, 0.5)
opti.solver("ipopt", {"print_time": False}, {"print_level": 0, "sb": "yes"})
solution = opti.solve()

print("stationary_point =", ca.DM(solution.value(probability)))
```
