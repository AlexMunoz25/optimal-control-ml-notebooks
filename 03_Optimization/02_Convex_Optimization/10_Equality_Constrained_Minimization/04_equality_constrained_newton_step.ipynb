### 3.2.10.4. Equality Constrained Newton Step


$$\begin{bmatrix}\nabla^2f(x)&A^T\\A&0\end{bmatrix}\begin{bmatrix}\Delta x\\w\end{bmatrix}=-\begin{bmatrix}\nabla f(x)\\0\end{bmatrix}$$


**Explanation:**

The equality-constrained Newton step minimizes the local quadratic model subject to the linearized equality constraint. Feasibility is the equality constraint or its linearized version; optimality is stationarity together with that feasibility condition.

The block matrix is the KKT system for the equality-constrained Newton step. The unknowns are the primal step $\Delta x$ and multiplier-like vector $w$, and the equations enforce stationarity and linearized feasibility together.

**Properties:**
- Equality constraints introduce unrestricted multipliers.
- KKT systems couple stationarity and feasibility.
- Eliminating constraints can reduce the number of variables.

**Intuition:**

The numerical example keeps the equality constraint visible in every algebraic step.


**Numerical Example:**

Let
$$
\nabla f=\begin{bmatrix}2\\4\end{bmatrix},\quad
H=\begin{bmatrix}2&0\\0&4\end{bmatrix},\quad
A=\begin{bmatrix}1&1\end{bmatrix}.
$$

The Newton equations are
$$
2\Delta x_1+w=-2,
\quad
4\Delta x_2+w=-4,
\quad
\Delta x_1+\Delta x_2=0.
$$

Solving gives
$$
\Delta x_1=1/3,
\qquad
\Delta x_2=-1/3.
$$

The step is feasible because its components sum to zero.


```python
import sympy as sp

step_1, step_2, multiplier = sp.symbols('Delta_x_1 Delta_x_2 w')
solution = sp.solve([sp.Eq(2 * step_1 + multiplier, -2), sp.Eq(4 * step_2 + multiplier, -4), sp.Eq(step_1 + step_2, 0)], [step_1, step_2, multiplier])

print("solution =", solution)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

kkt_matrix = ca.DM([[2, 0, 1], [0, 4, 1], [1, 1, 0]])
right_side = ca.DM([-2, -4, 0])
solution = ca.solve(kkt_matrix, right_side)

print("solution =", solution)
```
