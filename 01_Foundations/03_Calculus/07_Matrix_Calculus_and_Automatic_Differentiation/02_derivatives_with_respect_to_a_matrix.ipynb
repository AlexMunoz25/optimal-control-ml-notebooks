### 1.3.7.2. Derivatives with Respect to a Matrix

$$
\frac{\partial f}{\partial X} =
\begin{bmatrix} \dfrac{\partial f}{\partial X_{ij}} \end{bmatrix} \in \mathbb{R}^{m\times n},
\qquad
\frac{\partial}{\partial X}\operatorname{tr}(AX) = A^{\top} .
$$

**Explanation:**

The derivative of a scalar with respect to a matrix $X$ is the matrix of partials $\partial f/\partial X_{ij}$, laid out in the **same shape as $X$** so it can be used directly as a gradient-descent update on the matrix. This is the object optimizers and learning algorithms update when the unknown is itself a matrix — a weight matrix, a covariance, or a gain. The building-block identity $\partial\,\operatorname{tr}(AX)/\partial X = A^\top$ uses the [trace](../../01_Linear_Algebra/03_Matrix/06_matrix_transpose.ipynb) and recurs throughout matrix calculus.

**Properties:**
- $\dfrac{\partial}{\partial X}\operatorname{tr}(A^\top X) = A$, and $\dfrac{\partial}{\partial X}(\mathbf{a}^\top X \mathbf{b}) = \mathbf{a}\mathbf{b}^\top$.
- The gradient $\partial f/\partial X$ shares the shape of $X$, the rule that makes matrix gradient descent well defined.

**Numerical Example:**

Let $f(X) = \operatorname{tr}(AX)$ with $A = \begin{bmatrix} 2 & 1 \\ 0 & 3 \end{bmatrix}$ and $X = \begin{bmatrix} X_{11} & X_{12} \\ X_{21} & X_{22} \end{bmatrix}$. Expanding,

$$
\operatorname{tr}(AX) = 2X_{11} + 1\cdot X_{21} + 0\cdot X_{12} + 3X_{22},
$$

so the entrywise partials assemble into

$$
\frac{\partial f}{\partial X} =
\begin{bmatrix} 2 & 0 \\ 1 & 3 \end{bmatrix} = A^\top .
$$

```python
import sympy as sp

X_11, X_12, X_21, X_22 = sp.symbols("X_11 X_12 X_21 X_22")
X = sp.Matrix([[X_11, X_12], [X_21, X_22]])
A = sp.Matrix([[2, 1], [0, 3]])

scalar = (A * X).trace()
gradient_matrix = X.applyfunc(lambda entry: sp.diff(scalar, entry))

print("tr(AX) =", sp.expand(scalar))
print("∂ tr(AX)/∂X ="); sp.pprint(gradient_matrix)
print("equals Aᵀ:", gradient_matrix == A.T)
```

**Equivalent CasADi implementation:**

CasADi differentiates the scalar with respect to the vectorized matrix and reshapes the result back to $X$'s shape, recovering $A^\top$.

```python
import casadi as ca

X = ca.SX.sym("X", 2, 2)
A = ca.DM([[2, 1], [0, 3]])

scalar = ca.trace(A @ X)
gradient_matrix = ca.reshape(ca.gradient(scalar, ca.vec(X)), 2, 2)

print("∂ tr(AX)/∂X =", gradient_matrix)
print("Aᵀ          =", A.T)
```
