### 1.3.6.3. Change of Variables and the Jacobian Determinant

$$
\iint_{R} f(x, y)\,dA = \iint_{S} f\big(x(u,v), y(u,v)\big)\,
\left|\frac{\partial(x, y)}{\partial(u, v)}\right|\,du\,dv,
\qquad
\frac{\partial(x, y)}{\partial(u, v)} = \det J .
$$

**Explanation:**

Changing variables in a multiple integral replaces the area (or volume) element by the absolute value of the [Jacobian](../05_Multivariable_Differential_Calculus/07_jacobian.ipynb) [determinant](../../01_Linear_Algebra/03_Matrix/10_matrix_determinant.ipynb), which measures the local factor by which the transformation stretches area. This is the multivariable generalization of [u-substitution](../03_Single_Variable_Integration/04_integration_by_substitution.ipynb) (where the scalar $du = g'(x)\,dx$ plays the role of the Jacobian). The determinant correction is exactly the term that appears when transforming a probability density under an invertible map — the change-of-variables formula behind normalizing flows and reparameterization.

**Properties:**
- Polar coordinates $x = r\cos\theta,\ y = r\sin\theta$ have Jacobian $\dfrac{\partial(x,y)}{\partial(r,\theta)} = r$, so $dA = r\,dr\,d\theta$.
- For a density $p$, the transformed density carries the factor $\left|\det J\right|^{-1}$ (the inverse map's Jacobian).

**Intuition:**

The map $T$ carries a small rectangle in the $uv$-plane to a curved region in the $xy$-plane; the Jacobian determinant is the local area-scaling factor.

**Numerical Example:**

Find the area of the disk of radius $2$ using polar coordinates. The Jacobian determinant is

$$
\frac{\partial(x,y)}{\partial(r,\theta)} =
\det\begin{bmatrix} \cos\theta & -r\sin\theta \\ \sin\theta & r\cos\theta \end{bmatrix}
= r\cos^2\theta + r\sin^2\theta = r .
$$

So $dA = r\,dr\,d\theta$ and

$$
\text{Area} = \int_{0}^{2\pi}\!\int_{0}^{2} r\,dr\,d\theta
= \int_{0}^{2\pi} 2\,d\theta = 4\pi ,
$$

matching the familiar $\pi R^2 = \pi\cdot 2^2 = 4\pi$. The factor $r$ — not $1$ — is what makes the integral correct.

```python
import sympy as sp

r, theta = sp.symbols("r theta", positive=True)
x_map = r * sp.cos(theta)
y_map = r * sp.sin(theta)

jacobian = sp.Matrix([x_map, y_map]).jacobian([r, theta])
jacobian_determinant = sp.simplify(jacobian.det())
disk_area = sp.integrate(sp.integrate(jacobian_determinant, (r, 0, 2)), (theta, 0, 2 * sp.pi))

print("Jacobian ∂(x,y)/∂(r,θ) ="); sp.pprint(jacobian)
print("det J = |∂(x,y)/∂(r,θ)| =", jacobian_determinant)
print("area of disk (radius 2) =", disk_area, "= π R^2 =", sp.pi * 2**2)
```

**Equivalent CasADi implementation:**

CasADi forms the same polar Jacobian and its determinant from the coordinate map.

```python
import casadi as ca

r = ca.SX.sym("r")
theta = ca.SX.sym("theta")
coordinate_map = ca.vertcat(r * ca.cos(theta), r * ca.sin(theta))
jacobian = ca.jacobian(coordinate_map, ca.vertcat(r, theta))
determinant_function = ca.Function("determinant", [r, theta], [ca.det(jacobian)])

print("det J at (r=2, θ=π/4) =", float(determinant_function(2.0, ca.pi / 4)))
```
