### 1.1.6.1. Linear Transformation


$$
T(\alpha \mathbf{v}_1 + \beta \mathbf{v}_2) =
\alpha T(\mathbf{v}_1) + \beta T(\mathbf{v}_2)
$$


**Explanation:**

A linear transformation maps linear combinations of inputs to the same linear combinations of outputs. This preservation law is the central structural idea of linear algebra, and it leads directly to [image space](./02_image_space.ipynb) and [matrix representation](./05_matrix_representation.ipynb).

**Example:**

If

$$
T(x, y) =
\begin{bmatrix}
x \\
y \\
x + y
\end{bmatrix},
$$

then

$$
T(3, 4) =
\begin{bmatrix}
3 \\
4 \\
7
\end{bmatrix}.
$$


```python
import sympy as sp

input_vector = sp.Matrix([3, 4])
transformation_matrix = sp.Matrix([[1, 0], [0, 1], [1, 1]])
output_vector = transformation_matrix * input_vector

print("output_vector =", output_vector)
```
