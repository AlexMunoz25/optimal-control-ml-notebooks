### 1.3.1.4. Continuity

$$
f \text{ is continuous at } a \iff
\underbrace{f(a)\ \text{defined}}_{\text{(i)}}, \quad
\underbrace{\lim_{x\to a} f(x)\ \text{exists}}_{\text{(ii)}}, \quad
\underbrace{\lim_{x\to a} f(x) = f(a)}_{\text{(iii)}} .
$$

**Explanation:**

A function is continuous at a point $a$ when all three conditions hold: $f(a)$ is defined, the limit there exists, and the two coincide. If any condition fails the function is discontinuous at $a$ — a missing value or a value that disagrees with the limit gives a *removable* discontinuity, unequal one-sided limits give a *jump*, and an infinite limit gives an *infinite* discontinuity. A function is continuous on an interval if it is continuous at each point; polynomials are continuous everywhere and rational functions are continuous wherever the denominator is nonzero. Continuity is the hypothesis that powers the [intermediate value theorem](./06_intermediate_value_theorem.ipynb) and the [extreme value theorem](./07_extreme_value_theorem.ipynb), which guarantee that roots and optima exist.

**Properties:**
- Sums, products, quotients (nonzero denominator), and compositions of continuous functions are continuous.
- Continuity at $a$ is precisely the statement that the limit may be evaluated by direct substitution.

**Intuition:**

The three ways continuity can fail are a hole that can be filled (removable), a sudden jump between unequal one-sided limits, and a blow-up to infinity.

**Numerical Example:**

Compare two functions at $a = 2$.

For $f(x) = \dfrac{x^2 - 4}{x - 2}$, condition (i) fails: $f(2) = 0/0$ is undefined, so $f$ is **discontinuous** at $2$. The limit nonetheless exists, since for $x \neq 2$,

$$
\frac{x^2 - 4}{x - 2} = \frac{(x-2)(x+2)}{x-2} = x + 2 \;\Rightarrow\; \lim_{x\to 2} f(x) = 4 .
$$

Because the limit exists but the value is missing, the discontinuity is *removable* (define $f(2) := 4$ to repair it).

For $g(x) = x^2$, all three conditions hold at $2$: $g(2) = 4$, $\lim_{x\to 2} x^2 = 4$, and they agree, so $g$ is **continuous** at $2$.

```python
import sympy as sp

x = sp.symbols("x")
f = (x**2 - 4) / (x - 2)
g = x**2

f_value_defined = (x - 2).subs(x, 2) != 0
f_limit = sp.limit(f, x, 2)
f_simplified = sp.simplify(f)

g_value = g.subs(x, 2)
g_limit = sp.limit(g, x, 2)
g_continuous = g_limit == g_value

print("f(x)=(x^2-4)/(x-2):  f(2) defined:", f_value_defined, " limit:", f_limit, " simplifies to:", f_simplified)
print("  -> removable discontinuity at x=2 (limit exists, value missing)")
print("g(x)=x^2:  g(2) =", g_value, " limit:", g_limit, " continuous:", g_continuous)
```

```python
import numpy as np
import matplotlib.pyplot as plt

points = np.linspace(-1, 5, 400)
points = points[np.abs(points - 2) > 1e-6]
values = (points**2 - 4) / (points - 2)

plt.figure(figsize=(8, 5))
plt.plot(points, values, linewidth=2, label=r"$f(x)=(x^2-4)/(x-2)=x+2,\ x\neq 2$")
plt.scatter([2], [4], facecolors="none", edgecolors="red", s=120, zorder=5, label="removable hole at (2, 4)")
plt.xlabel("x"); plt.ylabel("f(x)")
plt.title("Removable discontinuity")
plt.legend(); plt.grid(True, alpha=0.3); plt.tight_layout()
plt.show()
```
