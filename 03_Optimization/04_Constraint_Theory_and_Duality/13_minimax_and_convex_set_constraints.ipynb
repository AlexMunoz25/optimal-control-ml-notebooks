### 3.4.13. Minimax and Convex Set Constraints

$$
\min_x \max\{f_1(x),\ldots,f_p(x)\}
\quad \Longleftrightarrow \quad
\min_{x,z} z \ \text{ subject to }\ f_j(x)\le z
$$

$$
\left(\mu_0^*\nabla f(x^*)+\sum_i\lambda_i^*\nabla h_i(x^*)+
\sum_j\mu_j^*\nabla g_j(x^*)\right)^T(x-x^*)\ge 0,\quad x\in X
$$

**Explanation:**

A minimax problem can be rewritten by introducing a scalar upper bound $z$ on all component functions. This turns the maximum into inequality constraints, so the multiplier theory developed for inequalities applies. If there is also a closed convex set constraint $x\in X$, the gradient stationarity equation becomes a variational inequality over $X$ rather than an unconstrained vector equation.

**Intuition:**

<img src="../../Figures/030413_berts_fig_3_3_3_convex_set_constraint.jpeg" alt="Bertsekas Figure 3.3.3: convex set constraints where ordinary Lagrange multipliers can fail" style="width: 60%; height: auto;">

The figure shows why a closed convex set can contribute its own normal geometry. The multiplier expression must support the set $X$ at the candidate point, not simply vanish as a vector.

**Numerical Example:**

Consider the one-dimensional minimax problem

$$
\min_x \max\{x, -x\}.
$$

Introduce $z$:

$$
\min_{x,z} z
\quad \text{subject to} \quad
x-z\le 0,
\qquad -x-z\le 0.
$$

At $x^*=0$, the smallest feasible $z$ is $z^*=0$ because

$$
x-z=0-0=0,
\qquad
-x-z=0-0=0.
$$

Both inequalities are active. The Lagrangian is

$$
L(x,z,\mu_1,\mu_2)=z+\mu_1(x-z)+\mu_2(-x-z).
$$

Stationarity gives

$$
\frac{\partial L}{\partial x}=\mu_1-\mu_2=0,
$$

and

$$
\frac{\partial L}{\partial z}=1-\mu_1-\mu_2=0.
$$

Thus

$$
\mu_1=\mu_2, \qquad 2\mu_1=1,
$$

so

$$
\mu_1^*=\mu_2^*=\frac{1}{2}.
$$

The multipliers are nonnegative and sum to one, matching the minimax multiplier condition for the active maximum terms.

```python
import sympy as sp

x, z, mu_1, mu_2 = sp.symbols('x z mu_1 mu_2')

lagrangian = z + mu_1 * (x - z) + mu_2 * (-x - z)
stationarity_equations = [
    sp.diff(lagrangian, x),
    sp.diff(lagrangian, z),
]
solution = sp.solve(stationarity_equations, [mu_1, mu_2], dict=True)[0]

result = {
    'solution': solution,
    'constraint_values_at_candidate': sp.Matrix([(x - z).subs({x: 0, z: 0}), (-x - z).subs({x: 0, z: 0})]),
    'stationarity_at_solution': sp.Matrix(stationarity_equations).subs(solution),
}

for label, value in result.items():
    print(f"{label} =", value)
```
