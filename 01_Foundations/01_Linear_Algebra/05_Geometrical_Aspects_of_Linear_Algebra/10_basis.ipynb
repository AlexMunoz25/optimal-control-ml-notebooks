### 1.1.5.10. Basis


$$
B = \left\{ \mathbf{b}_1, \mathbf{b}_2, \ldots, \mathbf{b}_n \right\}
$$

$$
\mathbf{v} = c_1 \mathbf{b}_1 + c_2 \mathbf{b}_2 + \cdots + c_n \mathbf{b}_n
$$


**Explanation:**

A basis is a linearly independent set of vectors that spans the whole space, so it acts as a coordinate frame for every vector in that space. Savov introduces the basis concept as the bridge between geometric vectors and their component descriptions, which is why this notebook comes immediately before the notebook on [coordinates with respect to a basis](./11_coordinates_with_respect_to_a_basis.ipynb).

<p align="center">
  <img src="../../../Figures/01010502_basis.png" alt="Alternative basis vectors" width="430">
</p>

**Properties:**
- In $\mathbb{R}^n$, every basis contains exactly $n$ vectors.
- Different bases describe the same vector space with different coordinates.

**Example:**

In $\mathbb{R}^2$,

$$
B =
\left\{
\begin{bmatrix}
1 \\
1
\end{bmatrix},
\begin{bmatrix}
1 \\
-1
\end{bmatrix}
\right\}
$$

is a basis because the two vectors are independent and span the whole plane.


```python
import sympy as sp

basis_matrix = sp.Matrix([[1, 1], [1, -1]])
determinant_value = basis_matrix.det()

print("basis_matrix =")
sp.pprint(basis_matrix)
print("determinant_value =", determinant_value)
print("is_basis =", determinant_value != 0)
```
