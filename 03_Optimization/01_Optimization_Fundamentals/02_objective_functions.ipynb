### 3.1.2. Objective Functions

$$
\min_{\mathbf{x}\in\mathbb{R}^n} \; f(\mathbf{x}),
\qquad f:\mathbb{R}^n\to\mathbb{R}.
$$

**Explanation:**

The objective function $f$ assigns a single real number to every choice of the [decision variables](./03_decision_variables.ipynb), and optimization searches for the choice that makes this number smallest (or largest). Minimizing $f$ and maximizing $-f$ describe the same problem, so a single convention — minimization — is enough. The value $f(\mathbf{x}^\star)$ at an optimal point is the *optimal value*, while any point attaining it is an *optimal solution*.

**Intuition:**

The objective can be pictured as a landscape over the [feasible set](./04_feasible_sets.ipynb): its level curves show equal-cost contours, and optimization moves toward the lowest contour the constraints allow.

![Objective over the feasible set](../../Figures/030101_jasour_feasible_set_objective.jpeg)

**Numerical Example:**

Consider the quadratic objective
$$
f(\mathbf{x}) = (x_1-1)^2 + 2x_2^2 + x_1 x_2.
$$

Sampling three candidate points,
$$
f(0,0) = 1,\qquad
f(1,0) = 0,\qquad
f(1,1) = 0 + 2 + 1 = 3.
$$

The gradient is
$$
\nabla f(\mathbf{x}) =
\begin{bmatrix} 2(x_1-1) + x_2 \\ 4x_2 + x_1 \end{bmatrix},
\qquad
\nabla f(1,0) = \begin{bmatrix}0\\1\end{bmatrix}\neq \mathbf 0,
$$
so $(1,0)$ lowers the cost below $(0,0)$ but is not yet stationary: a further decrease is possible by moving against the gradient.

```python
import sympy as sp

x_1, x_2 = sp.symbols("x_1 x_2")
decision_variable = sp.Matrix([x_1, x_2])
objective = (x_1 - 1)**2 + 2*x_2**2 + x_1*x_2

sample_points = [(0, 0), (1, 0), (1, 1)]
objective_values = [objective.subs({x_1: a, x_2: b}) for a, b in sample_points]
gradient = sp.Matrix([sp.diff(objective, variable) for variable in decision_variable])
gradient_at_candidate = gradient.subs({x_1: 1, x_2: 0})

print("sample objective values =", objective_values)
print("gradient ="); sp.pprint(gradient)
print("gradient at (1, 0) ="); sp.pprint(gradient_at_candidate)
```

**Equivalent casadi implementation:**

```python
import casadi as ca

decision_variable = ca.SX.sym("x", 2)
objective_expression = (decision_variable[0] - 1)**2 + 2*decision_variable[1]**2 + decision_variable[0]*decision_variable[1]
objective = ca.Function("f", [decision_variable], [objective_expression])
gradient = ca.Function("grad_f", [decision_variable], [ca.gradient(objective_expression, decision_variable)])

for point in ([0, 0], [1, 0], [1, 1]):
    print(f"f({point}) =", float(objective(point)))
print("gradient at (1, 0) =", gradient([1, 0]))
```
