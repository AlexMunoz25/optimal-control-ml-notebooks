### 1.1.9.10. Index Raising and Lowering

$$
g_{ij} A^j = A_i \qquad \text{(lowering)}, \qquad g^{ij} B_i = B^j \qquad \text{(raising)}
$$

$$
g^{ij} A_{ik} = A^j{}_k, \qquad g_{js} C^{is}{}_k = C^i{}_{jk}, \qquad g^{il} T^{jk}{}_l = T^{ijk}
$$

**Explanation:**

- Lowering contracts an upper index with the covariant metric $g_{ij}$.
- Raising contracts a lower index with the inverse metric $g^{ij}$.
- One metric contraction changes one slot at a time, so the same rule works for vectors and higher-rank tensors.
- In Cartesian coordinates the numerical values stay the same because $g_{ij} = \delta_{ij}$.

**Connections:**

- This turns the metric tensor from a geometric object into an algebraic tool.
- It is also the bridge between vectors and one-forms used in the next notebook.
- In optimization, preconditioning changes how gradient covectors are converted into update directions.

**Example:**

- In spherical coordinates,
$$
g_{ij} = \operatorname{diag}(1, r^2, r^2 \sin^2 \theta).
$$
- If $v^i = (\dot r, \dot \theta, \dot \phi)$, then
$$
v_i = (\dot r,\; r^2 \dot \theta,\; r^2 \sin^2 \theta\,\dot \phi).
$$


```python
import sympy as sp

radial_coordinate, polar_angle = sp.symbols('r theta', positive=True)
metric_tensor_covariant = sp.diag(
    1,
    radial_coordinate**2,
    radial_coordinate**2 * sp.sin(polar_angle)**2,
)
metric_tensor_contravariant = metric_tensor_covariant.inv()
contravariant_velocity = sp.Matrix([
    sp.Symbol('r_dot'),
    sp.Symbol('theta_dot'),
    sp.Symbol('phi_dot'),
])

covariant_velocity = metric_tensor_covariant * contravariant_velocity
raised_velocity = sp.simplify(metric_tensor_contravariant * covariant_velocity)

print("covariant_velocity =")
sp.pprint(covariant_velocity)
print("raised_velocity =")
sp.pprint(raised_velocity)
```
