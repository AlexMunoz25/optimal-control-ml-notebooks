### 1.1.7.13. Change of Basis to Eigenbasis


$$
\left[ \mathbf{v} \right]_{B_{\lambda}} = Q^{-1} \left[ \mathbf{v} \right]_{B_s}
$$


**Explanation:**

In an eigendecomposition, the matrix $Q$ is the change-of-basis matrix from the eigenbasis to the standard basis. Expressing the same vector in eigenbasis coordinates makes the matrix action diagonal, which is the interpretation built on top of [eigendecomposition](./04_eigendecomposition.ipynb).

**Example:**

If $Q$ contains the eigenvectors of $A$ as columns, then multiplying by $Q^{-1}$ converts a standard-basis coordinate vector into eigenbasis coordinates before the diagonal action of $\Lambda$ is applied.


```python
import sympy as sp

matrix_q = sp.Matrix([[1, 2], [2, -1]])
vector_in_standard_basis = sp.Matrix([3, 4])
vector_in_eigenbasis = sp.simplify(matrix_q.inv() * vector_in_standard_basis)

print("vector_in_eigenbasis =", vector_in_eigenbasis)
```
