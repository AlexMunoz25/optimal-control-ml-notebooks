### 1.1.3.15. Matrix Frobenius Norm

$$
\|\mathbf{A}\|_F =
\sqrt{\sum_{i=1}^m \sum_{j=1}^n |A_{ij}|^2}
$$

**Explanation:**

The Frobenius norm of a matrix $\mathbf{A} \in \mathbb{R}^{m \times n}$ is equivalent to the Euclidean norm of the vector formed by its entries.
It equals the square root of the sum of squares of the singular values and is invariant under unitary transformations.

**Related concepts:**
- Extends the [Norm of a Vector (Euclidean)](../02_Vector/06_vector_norm.ipynb) from vectors to matrices.
- It pairs naturally with [Trace of a Matrix](./17_matrix_trace.ipynb) and [Matrix Norm Inequality](./16_matrix_norm_inequality.ipynb).

**Example:**

Let

$$
\mathbf{A} =
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}.
$$

Then

$$
\|\mathbf{A}\|_F =
\sqrt{1^2 + 2^2 + 3^2 + 4^2}
=
\sqrt{30}.
$$


```python
import sympy as sp

matrix_a = sp.Matrix([[1, 2], [3, 4]])

frobenius_norm = matrix_a.norm()

print("‖A‖_F =", frobenius_norm)
```
