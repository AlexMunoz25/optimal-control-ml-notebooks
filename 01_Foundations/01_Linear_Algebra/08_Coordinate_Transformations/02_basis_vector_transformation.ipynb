### 1.1.8.2. Basis-Vector Transformation


$$
\hat{\mathbf{i}}' =
\begin{bmatrix}
\cos \alpha \\
\sin \alpha
\end{bmatrix},
\qquad
\hat{\mathbf{j}}' =
\begin{bmatrix}
-\sin \alpha \\
\cos \alpha
\end{bmatrix}
$$


**Explanation:**

A basis-vector transformation describes how the basis directions themselves move when the frame is rotated. In an active rotation, the basis vectors point in new directions relative to the original frame. This basis-side viewpoint is best understood alongside [passive](./03_passive_transformation.ipynb) and [active](./04_active_transformation.ipynb) interpretations.

**Example:**

Rotating the basis by $150^{\circ}$ replaces the original unit vectors with new unit vectors whose components in the original frame are given by the rotation matrix columns.


```python
import numpy as np

rotation_angle = np.deg2rad(150.0)
rotated_basis_i = np.array([np.cos(rotation_angle), np.sin(rotation_angle)])
rotated_basis_j = np.array([-np.sin(rotation_angle), np.cos(rotation_angle)])

print("rotated_basis_i =", np.round(rotated_basis_i, 4))
print("rotated_basis_j =", np.round(rotated_basis_j, 4))
```
