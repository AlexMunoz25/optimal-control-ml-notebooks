### 1.1.8.9. Einstein Summation Convention


$$
A^{i'} = \sum_{j=1}^{3} a_{ij} A^j = a_{ij} A^j,
\qquad
A_{i'} = \sum_{j=1}^{3} b_{ij} A_j = b_{ij} A_j
$$


**Explanation:**

Einstein summation convention suppresses the summation sign when an index appears twice in one term, once up and once down. The repeated index is a dummy index and is summed over, while the free index labels the output component. This notation turns long component-wise transformation rules into compact expressions without changing the underlying algebra.

**Example:**

Let

$$
A=\begin{bmatrix}2&-1&0\\1&3&4\\0&2&1\end{bmatrix},
\qquad
\mathbf{x}=\begin{bmatrix}1\\2\\-1\end{bmatrix}.
$$

Einstein summation says

$$
y^i=A^i{}_j x^j.
$$

Compute the components explicitly:

$$
y^1=2(1)+(-1)(2)+0(-1)=0,
$$

$$
y^2=1(1)+3(2)+4(-1)=1+6-4=3,
$$

and

$$
y^3=0(1)+2(2)+1(-1)=3.
$$

So

$$
\mathbf{y}=\begin{bmatrix}0\\3\\3\end{bmatrix}.
$$


```python
import sympy as sp

transformation_matrix = sp.Matrix([
    [2, -1, 0],
    [1, 3, 4],
    [0, 2, 1],
])
contravariant_components = sp.Matrix([1, 2, -1])
transformed_components_from_matrix_product = transformation_matrix * contravariant_components
transformed_components_from_explicit_sums = sp.Matrix([
    sum(transformation_matrix[row, column] * contravariant_components[column]
        for column in range(transformation_matrix.cols))
    for row in range(transformation_matrix.rows)
])

print("transformed_components_from_matrix_product =")
sp.pprint(transformed_components_from_matrix_product)
print("transformed_components_from_explicit_sums =")
sp.pprint(transformed_components_from_explicit_sums)
```
