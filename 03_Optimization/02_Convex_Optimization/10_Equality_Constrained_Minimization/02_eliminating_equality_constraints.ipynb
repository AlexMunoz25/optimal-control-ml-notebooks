### 3.2.10.2. Eliminating Equality Constraints


$$x=Fz+x_0,
\qquad AF=0,
\qquad Ax_0=b$$


**Explanation:**

Eliminating equality constraints rewrites every feasible point as a particular solution plus a nullspace direction. The reduced problem is unconstrained in fewer variables. Feasibility is the equality constraint or its linearized version; optimality is stationarity together with that feasibility condition.

The vector $x_0$ is one particular solution of $Ax=b$. The columns of $F$ lie in the nullspace of $A$, so $x=Fz+x_0$ describes all feasible points by changing only the free coordinate vector $z$.

**Properties:**
- Equality constraints introduce unrestricted multipliers.
- KKT systems couple stationarity and feasibility.
- Eliminating constraints can reduce the number of variables.

**Intuition:**

The numerical example keeps the equality constraint visible in every algebraic step.


**Numerical Example:**

Use the equality constraint
$$
x_1+x_2=2.
$$

Set
$$
x_1=z,\qquad x_2=2-z.
$$

For the objective $\frac12(x_1^2+x_2^2)$, the reduced objective is
$$
\frac12\left(z^2+(2-z)^2\right)=z^2-2z+2.
$$

Differentiate:
$$
2z-2=0\quad\Longrightarrow\quad z=1.
$$

Thus $x_1=1$ and $x_2=1$.


```python
import sympy as sp

reduced_variable = sp.symbols('z')
reduced_objective = sp.Rational(1, 2) * (reduced_variable ** 2 + (2 - reduced_variable) ** 2)
optimizer = sp.solve(sp.diff(reduced_objective, reduced_variable), reduced_variable)[0]
solution = sp.Matrix([optimizer, 2 - optimizer])

print("solution =", solution)
```

**Direct `casadi` implementation:**

```python
import casadi as ca

reduced_variable = ca.SX.sym("z")
reduced_objective = 0.5 * (reduced_variable ** 2 + (2 - reduced_variable) ** 2)
gradient = ca.gradient(reduced_objective, reduced_variable)
hessian = ca.jacobian(gradient, reduced_variable)
optimizer = ca.solve(ca.evalf(hessian), -ca.evalf(ca.substitute(gradient, reduced_variable, ca.DM(0))))
solution = ca.vertcat(optimizer, 2 - optimizer)

print("solution =", solution)
```
