### 3.1.26. Finite Difference Derivative Approximations


$$
\frac{\partial f(x)}{\partial x_i}
\approx \frac{f(x+he_i)-f(x)}{h},
\qquad
\frac{\partial f(x)}{\partial x_i}
\approx \frac{f(x+he_i)-f(x-he_i)}{2h}.
$$


**Explanation:**

Nonderivative methods begin with the practical fact that derivatives may be unavailable, unreliable, or too expensive to derive analytically. Finite differences approximate derivatives using only function evaluations. Forward differences are cheaper but have first-order truncation error, while central differences require more function evaluations but have second-order truncation error in exact arithmetic. It is also important that $h$ cannot simply be made arbitrarily small, because roundoff error grows when nearly equal function values are subtracted.

The vector $e_i$ is the coordinate basis direction for the $i$th variable. The forward formula compares $f$ at $x$ and $x+he_i$, while the centered formula compares symmetric perturbations $x\pm he_i$ to approximate the same partial derivative.

This topic is important for engineering computation because many objectives come from simulations or experiments. In such cases, one may know how to evaluate the cost but not have an exact formula for its gradient. Finite differences provide a bridge to gradient and Newton-like methods, but the choice of perturbation size and scaling becomes part of the numerical method.

**Intuition:**

Near a stationary point, the true derivative may be small, so the relative error in a finite-difference approximation can become large. The code cell compares forward and central differences for a smooth scalar function at one point. The central approximation is typically more accurate for the same $h$, but it costs an additional function evaluation.


**Numerical Example:**

Approximate the derivative of $f(x)=\sin x$ at $x=1$ with $h=0.001$. The exact derivative is
$$
f'(1)=\cos(1)\approx 0.540302.
$$

The forward difference uses $f(1+h)$ and $f(1)$:
$$
\frac{\sin(1.001)-\sin(1)}{0.001}
\approx \frac{0.842011-0.841471}{0.001}
\approx 0.539881.
$$

The central difference uses both sides of the point:
$$
\frac{\sin(1.001)-\sin(0.999)}{2(0.001)}
\approx \frac{0.842011-0.840930}{0.002}
\approx 0.540302.
$$

For this smooth function and stepsize, the central difference is much closer to the exact derivative.


```python
import numpy as np

point = 1.0
stepsize = 1e-3
objective = np.sin
forward_difference = (objective(point + stepsize) - objective(point)) / stepsize
central_difference = (objective(point + stepsize) - objective(point - stepsize)) / (2 * stepsize)
exact_derivative = np.cos(point)

forward_difference, central_difference, exact_derivative
```
