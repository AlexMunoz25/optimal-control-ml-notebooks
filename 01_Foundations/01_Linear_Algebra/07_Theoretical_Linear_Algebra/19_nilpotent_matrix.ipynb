### 1.1.7.19. Nilpotent Matrix


$$
N^k = \mathbf{0} \quad \text{for some positive integer } k
$$


**Explanation:**

A nilpotent matrix becomes the zero matrix under repeated self-multiplication. The smallest positive integer $k$ for which $N^k = \mathbf{0}$ is the index of nilpotency. Nilpotent matrices always have all eigenvalues equal to zero, which forces their trace and determinant to be zero as well. Strictly upper or lower triangular matrices are the canonical examples: for an $n \times n$ strictly triangular matrix the index is at most $n$. Nilpotency also connects to the [idempotent projection matrices](./12_projection_matrix_as_idempotent.ipynb) through the complementary condition $\Pi^2 = \Pi$ versus $N^k = \mathbf{0}$.

**Properties:**

- All eigenvalues are zero.
- Trace $= 0$ and determinant $= 0$.
- For an $n \times n$ nilpotent matrix, $N^n = \mathbf{0}$ always holds.
- $N$ is nilpotent if and only if its characteristic polynomial is $\lambda^n$.
- The matrix exponential $e^N = I + N + \tfrac{1}{2!}N^2 + \cdots$ is a finite sum.

**Example:**

The matrix

$$
N =
\begin{bmatrix}
-2 & 4 \\
-1 & 2
\end{bmatrix}
$$

satisfies

$$
N^2 =
\begin{bmatrix}
(-2)(-2) + (4)(-1) & (-2)(4) + (4)(2) \\
(-1)(-2) + (2)(-1) & (-1)(4) + (2)(2)
\end{bmatrix}
=
\begin{bmatrix} 0 & 0 \\ 0 & 0 \end{bmatrix},
$$

so $N$ is nilpotent with index $k = 2$.


```python
import sympy as sp

matrix_n = sp.Matrix([[-2, 4], [-1, 2]])

print("N^2 =")
sp.pprint(matrix_n ** 2)
print("eigenvalues =", list(matrix_n.eigenvals().keys()))
print("trace      =", matrix_n.trace())
print("determinant =", matrix_n.det())
```
