### 1.1.3.3. Matrix Scalar Multiplication

$$
\alpha \mathbf{A} = \alpha 
\begin{bmatrix}
a_{11} & a_{12} \\
a_{21} & a_{22}
\end{bmatrix}
=
\begin{bmatrix}
\alpha a_{11} & \alpha a_{12} \\
\alpha a_{21} & \alpha a_{22}
\end{bmatrix}
$$

**Explanation:**

Scaling a matrix by a scalar is useful in ML for adjusting learning rates or normalization.

**Related concepts:**
- Builds on [Matrices](./01_matrix_definition.ipynb).
- Combines with [Matrix Addition](./02_matrix_addition.ipynb) and supports later linear transformations in [Matrix-Vector Multiplication](./04_matrix_vector_multiplication.ipynb).

**Example:**

If $\alpha = 2$ and

$$
\mathbf{A} =
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix},
$$

then

$$
\alpha \mathbf{A} =
\begin{bmatrix}
2 & 4 \\
6 & 8
\end{bmatrix}.
$$


```python
import sympy as sp

scalar = 2
matrix_a = sp.Matrix([[1, 2], [3, 4]])

result = scalar * matrix_a

print("αA =")
sp.pprint(result)
```
