### 3.1.18. Incremental Gradient Methods


$$
f(x)=\sum_{i=1}^m f_i(x),
\qquad
x_{i+1}^k = x_i^k - \alpha^k \nabla f_i(x_i^k),
\qquad
x^{k+1}=x_m^k.
$$


**Explanation:**

Incremental gradient methods exploit objectives that are sums of many component functions. Instead of computing the full gradient of $f=\sum_i f_i$ before every update, the method processes one component or data block at a time. This can reduce the cost of each update and can produce useful progress before a full pass through all data has been completed. Bertsekas discusses these methods in the least-squares section because large data-fitting problems naturally decompose into residual blocks.

The price of incrementalism is that the direction at an intermediate substep is not the true full gradient direction. Far from the solution, this can be beneficial because each component quickly pulls the iterate toward satisfying one part of the model. Near the solution, however, the analysis becomes more delicate because component gradients may not vanish simultaneously at every substep.

**Example:**

For scalar linear fitting, each data point defines one residual term and one component gradient. A cycle of incremental updates visits these terms one at a time, adjusting the parameter after each residual. The code cell shows one such pass through four data points, illustrating the mechanics before any convergence theorem is invoked.


```python
import numpy as np

features = np.array([1.0, 2.0, 3.0, 4.0])
observations = np.array([1.1, 2.1, 2.9, 4.2])
parameter = 0.0
stepsize = 0.05

for feature, observation in zip(features, observations):
    residual = parameter * feature - observation
    parameter = parameter - stepsize * residual * feature

parameter
```
