### 1.1.9.10. Index Raising and Lowering

$$
A_i = g_{ij} A^j,
\qquad
B^j = g^{ij} B_i,
\qquad
A^j{}_k = g^{ij} A_{ik}
$$

**Explanation:**

The metric tensor does more than measure lengths: it also converts upper indices to lower indices and lower indices to upper ones. This gives an algebraic route between the descriptions introduced in [Contravariant Components](../08_Coordinate_Transformations/07_contravariant_components.ipynb) and [Covariant Components](../08_Coordinate_Transformations/08_covariant_components.ipynb). In Cartesian orthonormal coordinates the numerical values stay the same, but in curvilinear coordinates the metric generally changes them.

**Example:**

In spherical coordinates,
$$
g_{ij} = \operatorname{diag}(1, r^2, r^2 \sin^2 \theta).
$$
If
$$
v^i = (\dot{r}, \dot{\theta}, \dot{\phi}),
$$
then lowering the index gives
$$
v_i = (\dot{r}, r^2 \dot{\theta}, r^2 \sin^2 \theta\,\dot{\phi}).
$$

```python
import sympy as sp

radial_coordinate, polar_angle = sp.symbols('r theta', positive=True)
radial_velocity, polar_velocity, azimuthal_velocity = sp.symbols(
    'radial_velocity polar_velocity azimuthal_velocity'
)
metric_tensor_covariant = sp.diag(
    1,
    radial_coordinate**2,
    radial_coordinate**2 * sp.sin(polar_angle)**2,
)
metric_tensor_contravariant = metric_tensor_covariant.inv()
contravariant_velocity = sp.Matrix(
    [radial_velocity, polar_velocity, azimuthal_velocity]
)

covariant_velocity = metric_tensor_covariant * contravariant_velocity
raised_velocity = sp.simplify(metric_tensor_contravariant * covariant_velocity)

print("covariant_velocity =")
sp.pprint(covariant_velocity)
print("raised_velocity =")
sp.pprint(raised_velocity)
```
