### 3.2.5.7. Perturbation Sensitivity and Shadow Prices


$$p^*(u)\geq p^*(0)-\lambda^{*T}u$$


**Explanation:**

Perturbation analysis studies how the optimal value changes when constraint right-hand sides are shifted. Dual variables act as shadow prices because they define supporting slopes for the value function.

The vector $u$ perturbs constraint right-hand sides, and $\lambda^*$ is the optimal multiplier vector. The [inner product](../../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/02_inner_product_space.ipynb) $\lambda^{*T}u$ gives the first-order shadow-price correction to the optimal value bound.

**Properties:**
- Weak duality gives a lower bound for minimization.
- A primal feasible point gives an upper bound.
- A zero gap proves optimality when both sides are feasible.

**Intuition:**

The figure above gives the certificate geometry; the arithmetic computes the gap between the certificates.

<img src="../../../Figures/03020507_bv_fig_5_10_sensitivity_shadow_price.jpeg" alt="Boyd and Vandenberghe Figure 5.10: optimal value as a function of perturbation" style="width: 60%; height: auto;">


**Numerical Example:**

The figure above gives the certificate geometry; the arithmetic computes the gap between the certificates.

Suppose a primal feasible point has value $p=4$ and a dual feasible point has value $d=3$. Weak duality gives
$$
d\leq p^\star\leq p.
$$

Substitute the values:
$$
3\leq p^\star\leq4.
$$

The certificate gap is
$$
p-d=4-3=1.
$$

The dual point proves that no feasible primal solution can have value below $3$, while the primal point proves the optimum is at most $4$.


```python
import sympy as sp

primal_value = sp.Integer(4)
bound_value = sp.Integer(3)
gap = primal_value - bound_value

bound_value <= primal_value, gap
```
