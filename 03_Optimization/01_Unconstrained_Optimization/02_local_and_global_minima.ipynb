### 3.1.2. Local and Global Minima


$$
\begin{aligned}
x^* \in \arg\min_{x \in \mathbb{R}^n} f(x)
&\Longleftrightarrow
f(x^*) \le f(x), \quad \forall x \in \mathbb{R}^n, \\
x^* \text{ is local}
&\Longleftrightarrow
\exists \epsilon > 0:\ f(x^*) \le f(x),\quad \forall x \text{ with } \|x-x^*\| < \epsilon.
\end{aligned}
$$


**Explanation:**

In an unconstrained problem, there are no feasibility restrictions other than belonging to the ambient space $\mathbb{R}^n$. This makes the meaning of “minimum” especially important: a global minimizer is compared with every point in the space, while a local minimizer is compared only with points in a sufficiently small neighborhood. This distinction comes first because most algorithms in nonlinear programming see only local information, such as nearby function values, gradients, or local quadratic models. A neighborhood is measured by a [vector norm](../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb), so the basic language of local optimality already depends on the geometry introduced in linear algebra.

The first line compares $x^*$ with every point in the ambient [vector space](../../01_Foundations/01_Linear_Algebra/05_Geometrical_Aspects_of_Linear_Algebra/13_vector_space.ipynb), so it defines a global minimizer. The second line restricts the comparison to points with small [norm](../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb) distance from $x^*$, so it defines a local minimizer.

For a first course, it helps to read the definition as a warning. A point may look best from nearby but still be worse than another point far away, especially when the cost is nonconvex. This is why later optimality conditions are local unless convexity or some other global structure is available.

**Intuition:**

The figure shows a one-dimensional cost with several valleys and peaks. It is possible for a point to be locally minimal because every nearby perturbation raises the cost, while another valley gives a smaller value and is therefore the global minimum. The code cell computes the same sampled comparison used in the numerical example.

**Numerical Example:**

Using the one-dimensional valley picture in the figure, sample a cost curve at five points:
$$
\begin{array}{c|ccccc}
x & -2 & -1 & 0 & 1 & 2 \\
\hline
f(x) & 3 & 1 & 4 & 2 & 0
\end{array}
$$

At $x=-1$, the nearby sampled values are larger:
$$
f(-2)=3>1=f(-1),\qquad f(0)=4>1=f(-1).
$$

So $x=-1$ behaves like a local minimum in that valley. But
$$
f(2)=0<1=f(-1),
$$
so $x=-1$ is not global; among these samples, $x=2$ gives the global minimum value.


```python
import numpy as np

sample_points = np.array([-2.0, -1.0, 0.0, 1.0, 2.0])
objective_values = np.array([3.0, 1.0, 4.0, 2.0, 0.0])
local_minimum_index = 1
global_minimum_index = np.argmin(objective_values)
local_minimum = sample_points[local_minimum_index]
global_minimum = sample_points[global_minimum_index]
local_minimum_value = objective_values[local_minimum_index]
global_minimum_value = objective_values[global_minimum_index]

local_minimum, local_minimum_value, global_minimum, global_minimum_value
```
