### 3.3.8. Perspective and Linear-Fractional Maps

$$P(z,t)=z/t,\qquad t>0$$


**Explanation:**

The perspective map normalizes coordinates by a positive scale. Linear-fractional maps combine an [affine transformation](../../01_Foundations/01_Linear_Algebra/06_Linear_Transformations/08_affine_transformation.ipynb) with this normalization, which explains why many ratio expressions become convex-set transformations after lifting.

The map $P(z,t)$ takes a [vector](../../01_Foundations/01_Linear_Algebra/02_Vector/01_vector_definition.ipynb) $z$ and a positive scalar $t$. Dividing by $t$ normalizes every coordinate of $z$, so the positivity condition $t>0$ is part of the domain.

**Properties:**
- The scaling coordinate must be positive.
- The map sends rays to normalized coordinates.
- Linear-fractional maps combine affine maps with this normalization.

**Intuition:**

The figure above gives the geometric normalization; the arithmetic performs the same normalization numerically.

**Numerical Example:**

The figure above gives the geometric normalization; the arithmetic performs the same normalization numerically.

Let
$$
z=\begin{bmatrix}6\\3\end{bmatrix},\quad t=3>0.
$$

The perspective image is
$$
P(z,t)=z/t.
$$

Compute each coordinate:
$$
\frac{z}{t}=\frac13\begin{bmatrix}6\\3\end{bmatrix}=\begin{bmatrix}2\\1\end{bmatrix}.
$$

The positive scale keeps the transformation in its domain.


```python
import sympy as sp

point = sp.Matrix([6, 3])
scale = sp.Integer(3)
image = point / scale

print("image =", image)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

point = ca.DM([6, 3])
scale = 3
image = point / scale

print("image =", image)
```
