### 3.2.3.5. Sublevel Sets and Epigraphs


$$C_\alpha=\{x\mid f(x)\leq\alpha\},\qquad\operatorname{epi}f=\{(x,t)\mid f(x)\leq t\}$$


**Explanation:**

Sublevel sets and epigraphs translate function inequalities into geometry. A sublevel set keeps points below a specified value, while an epigraph records all points above the graph of the function.

**Properties:**
- Convex functions have convex epigraphs.
- Quasiconvex functions have convex sublevel sets.
- Epigraph transformations replace objectives by constraints.

**Intuition:**

The figure above shows set membership derived from a function value; the arithmetic checks that membership.

<img src="../../../Figures/03020305_bv_fig_3_5_epigraph.jpeg" alt="Boyd and Vandenberghe Figure 3.5: epigraph of a function" style="width: 60%; height: auto;">


**Numerical Example:**

The figure above shows set membership derived from a function value; the arithmetic checks that membership.

Let $f(x)=x^2$ and $x=2$. First compute
$$
f(2)=2^2=4.
$$

At level $5$,
$$
4\leq5,
$$
so $x=2$ is in the sublevel set. At level $3$,
$$
4\leq3
$$
is false, so $x=2$ is not in that lower sublevel set.


```python
import sympy as sp

function_value = sp.Integer(2) ** 2

function_value <= 5, function_value <= 3
```
