### 3.1.16. Least Squares Problem Models


$$
\min_{x\in\mathbb{R}^n} f(x),
\qquad
f(x)=\frac{1}{2}\|g(x)\|^2
=\frac{1}{2}\sum_{i=1}^m g_i(x)^2.
$$


**Explanation:**

Least squares problems measure the size of a residual vector. They appear whenever a model is chosen so that predicted outputs match data, measurements, or desired behavior as closely as possible. Bertsekas presents curve fitting, dynamic system identification, neural networks, and pattern classification as different instances of the same mathematical form. The residual norm is a [vector norm](../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb), and in the linear case the problem connects directly to projections and normal equations.

For students new to optimization, least squares is a useful bridge between linear algebra and nonlinear programming. The familiar linear least squares problem becomes nonlinear when the residuals depend nonlinearly on the parameters. The objective still has the same sum-of-squares structure, but the optimization landscape can now be curved or nonconvex.

**Example:**

The Bertsekas figure shows a least squares cost surface for a neural-network example, emphasizing that even a sum of squares can have a nontrivial shape. The code cell solves a small linear fitting problem so that the basic residual-minimization idea is visible before moving to nonlinear Gauss-Newton steps.

<img src="../../Figures/030101_berts_fig_1_5_1_least_squares_surface.jpeg" alt="Bertsekas Figure 1.5.1: least squares cost surface" style="width: 60%; height: auto;">


```python
import numpy as np

sample_points = np.array([0.0, 1.0, 2.0, 3.0])
observations = np.array([1.0, 2.0, 2.9, 4.2])
design_matrix = np.column_stack([sample_points, np.ones_like(sample_points)])
parameters, residuals, rank, singular_values = np.linalg.lstsq(design_matrix, observations, rcond=None)

parameters, residuals
```
