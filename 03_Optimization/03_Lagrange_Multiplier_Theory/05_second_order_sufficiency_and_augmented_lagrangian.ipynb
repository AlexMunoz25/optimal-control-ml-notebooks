### 3.3.5. Second-Order Sufficiency and Augmented Lagrangian

$$
y^T\nabla_{xx}^2L(x^*,\lambda^*)y>0
\quad \text{for all } y\neq 0 \text{ with } \nabla h(x^*)^T y=0
$$

$$
L_c(x,\lambda)=f(x)+\lambda^T h(x)+\frac{c}{2}\|h(x)\|^2
$$

**Explanation:**

Second-order sufficiency checks the Lagrangian curvature only along first-order feasible variations, not in every direction. The augmented Lagrangian adds a squared constraint violation term so that sufficiently large $c$ supplies [positive definite](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/17_positive_definite_matrix.ipynb) curvature in directions normal to the constraint surface. In this way, a constrained strict local minimum can become an unconstrained local minimum of $L_c(\cdot,\lambda^*)$.

**Intuition:**

<img src="../../Figures/03030305_berts_fig_3_2_1_augmented_curvature_lemma.jpeg" alt="Bertsekas Figure 3.2.1: adding scaled positive semidefinite curvature to make a matrix positive definite" style="width: 60%; height: auto;">

<img src="../../Figures/03030305_berts_fig_3_2_2_augmented_lagrangian_minimum.jpeg" alt="Bertsekas Figure 3.2.2: augmented Lagrangian contours for an equality-constrained minimum" style="width: 60%; height: auto;">

The first figure shows the matrix idea: add enough positive curvature in the directions where the constraint changes. The second figure shows the same idea for $x_2=0$: the penalty term turns negative normal curvature into positive curvature when $c$ is large enough.

**Numerical Example:**

Use the example in the second figure:

$$
f(x)=\frac{1}{2}(x_1^2-x_2^2)-x_2,
\qquad h(x)=x_2.
$$

At $x^*=(0,0)$, feasibility gives $h(x^*)=0$. The Lagrangian is

$$
L(x,\lambda)=\frac{1}{2}(x_1^2-x_2^2)-x_2+\lambda x_2.
$$

Stationarity at $x^*$ gives

$$
\nabla_x L(x^*,\lambda)=
\begin{bmatrix}0\\-1+\lambda\end{bmatrix}=0,
$$

so

$$
\lambda^*=1.
$$

The augmented Lagrangian is

$$
L_c(x,1)=\frac{1}{2}(x_1^2-x_2^2)-x_2+x_2+\frac{c}{2}x_2^2
=\frac{1}{2}x_1^2+\frac{c-1}{2}x_2^2.
$$

For the concrete value $c=2$ shown by the augmented contours,

$$
L_2(x,1)=\frac{1}{2}x_1^2+\frac{1}{2}x_2^2,
$$

which has positive curvature in both coordinate directions and an unconstrained minimum at $(0,0)$.

```python
import sympy as sp

x_1, x_2, lambda_multiplier, penalty = sp.symbols('x_1 x_2 lambda c', positive=True)

cost = sp.Rational(1, 2) * (x_1**2 - x_2**2) - x_2
constraint = x_2
lagrangian = cost + lambda_multiplier * constraint
stationarity = sp.Matrix([sp.diff(lagrangian, x_1), sp.diff(lagrangian, x_2)]).subs({x_1: 0, x_2: 0})
lambda_solution = sp.solve(stationarity, [lambda_multiplier], dict=True)[0]
augmented_lagrangian = sp.simplify(lagrangian.subs(lambda_solution) + penalty * constraint**2 / 2)
hessian_augmented = sp.hessian(augmented_lagrangian, (x_1, x_2))

result = {
    'lambda_solution': lambda_solution,
    'augmented_lagrangian_c_2': augmented_lagrangian.subs(penalty, 2),
    'hessian_augmented_c_2': hessian_augmented.subs(penalty, 2),
}
result
```
