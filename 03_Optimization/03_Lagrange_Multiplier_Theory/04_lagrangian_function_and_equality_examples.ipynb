### 3.3.4. Lagrangian Function and Equality Examples

$$
L(x,\lambda)=f(x)+\sum_{i=1}^{m}\lambda_i h_i(x)
$$

$$
\nabla_x L(x^*,\lambda^*)=0, \qquad \nabla_\lambda L(x^*,\lambda^*)=h(x^*)=0
$$

**Explanation:**

The Lagrangian function packages the objective and equality constraints into one expression. The equation $\nabla_x L=0$ says that objective and constraint gradients balance at the candidate point, while $\nabla_\lambda L=0$ simply restates feasibility. This notation is central because the same object later supports second-order conditions, sensitivity analysis, and duality.

**Properties:**
- Multipliers are variables in the Lagrangian, but feasibility is still enforced through $h(x^*)=0$.
- The [Hessian](../../README.md#missing-prerequisite-notes) of the Lagrangian, restricted to feasible first-order variations, carries second-order curvature information.

**Numerical Example:**

Use the equality example

$$
\min \frac{1}{2}(x_1^2+x_2^2+x_3^2)
\quad \text{subject to} \quad x_1+x_2+x_3=3.
$$

The Lagrangian is

$$
L(x,\lambda)=\frac{1}{2}(x_1^2+x_2^2+x_3^2)+\lambda(x_1+x_2+x_3-3).
$$

Stationarity gives

$$
\frac{\partial L}{\partial x_1}=x_1+\lambda=0,
\quad
\frac{\partial L}{\partial x_2}=x_2+\lambda=0,
\quad
\frac{\partial L}{\partial x_3}=x_3+\lambda=0.
$$

Thus

$$
x_1=x_2=x_3=-\lambda.
$$

Feasibility gives

$$
-\lambda-\lambda-\lambda=3
\quad \Rightarrow \quad
\lambda=-1.
$$

Therefore

$$
x^*=(1,1,1), \qquad \lambda^*=-1.
$$

The Lagrangian stationarity equations and the original equality constraint together determine both the point and the multiplier.

```python
import sympy as sp

x_1, x_2, x_3, lambda_multiplier = sp.symbols('x_1 x_2 x_3 lambda')

cost = sp.Rational(1, 2) * (x_1**2 + x_2**2 + x_3**2)
constraint = x_1 + x_2 + x_3 - 3
lagrangian = cost + lambda_multiplier * constraint
stationarity_equations = [
    sp.diff(lagrangian, x_1),
    sp.diff(lagrangian, x_2),
    sp.diff(lagrangian, x_3),
    sp.diff(lagrangian, lambda_multiplier),
]
solution = sp.solve(stationarity_equations, [x_1, x_2, x_3, lambda_multiplier], dict=True)[0]

result = {
    'lagrangian': lagrangian,
    'solution': solution,
    'cost_at_solution': cost.subs(solution),
}
result
```
