### 3.2.8.3. Distance Between Convex Sets


$$\operatorname{dist}(C,D)=\inf\{\|x-y\|_2\mid x\in C,
 y\in D\}$$


**Explanation:**

The distance between two convex sets is the smallest norm between one point from each set. When the sets are disjoint, the closest pair provides separation geometry. Feasibility requires choosing one point in each set; optimality minimizes their separation.

The formula searches over pairs of points with $x\in C$ and $y\in D$. The infimum of the Euclidean [norm](../../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb) $\|x-y\|_2$ is the distance between the two sets.

**Properties:**
- Distance is nonnegative.
- A positive distance indicates disjoint sets.
- Closest points often define a separating normal.

**Intuition:**

The figure above shows closest sets; the calculation computes the closest endpoints of two intervals.

<img src="../../../Figures/03020803_bv_fig_8_2_distance_between_polyhedra.jpeg" alt="Boyd and Vandenberghe Figure 8.2: Euclidean distance between polyhedra" style="width: 60%; height: auto;">


**Numerical Example:**

The figure above shows set-to-set distance; this scalar calculation uses intervals
$$
C=[-1,1],\qquad D=[3,5].
$$

The closest point in $C$ is $1$, and the closest point in $D$ is $3$. Hence
$$
\operatorname{dist}(C,D)=|3-1|=2.
$$

The positive distance shows that the two sets are disjoint.


```python
import sympy as sp

closest_point_c = sp.Integer(1)
closest_point_d = sp.Integer(3)
distance = abs(closest_point_d - closest_point_c)

distance
```
