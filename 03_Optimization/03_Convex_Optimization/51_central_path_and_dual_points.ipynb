### 3.3.51. Central Path and Dual Points

$$x^*(t)=\arg\min_x tf_0(x)+\phi(x)$$


**Explanation:**

The central path is the set of exact minimizers of barrier subproblems as the barrier parameter changes. Larger parameter values put more weight on the original objective and less relative weight on the barrier. Feasibility stays in the strict interior, and optimality improves as the barrier gap decreases.

For a fixed positive parameter $t$, the point $x^*(t)$ minimizes the barrier objective $tf_0(x)+\phi(x)$. As $t$ changes, these minimizers [trace](../../01_Foundations/01_Linear_Algebra/03_Matrix/17_matrix_trace.ipynb) the central path.

**Properties:**
- Each parameter defines a centering problem.
- The central path remains strictly feasible.
- Dual points can be recovered from the barrier optimality equations.

**Numerical Example:**

The figure above shows central-path movement; the arithmetic solves one point on the path.

For the one-dimensional centering problem
$$
\min_{x>0} tx-\log x,
$$
the derivative is
$$
t-\frac1x.
$$

The central point satisfies
$$
t-\frac1x=0.
$$

Thus
$$
tx=1 \quad\Longrightarrow\quad x^\star(t)=\frac1t.
$$

For $t=4$, the central point is $x^\star(4)=1/4$.


```python
import sympy as sp

t_parameter = sp.Integer(4)
central_point = 1 / t_parameter

print("central_point =", central_point)
```
