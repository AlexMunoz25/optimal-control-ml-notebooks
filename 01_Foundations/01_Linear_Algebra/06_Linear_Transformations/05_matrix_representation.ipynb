### 1.1.6.5. Matrix Representation of a Linear Transformation


$$
M_T =
\begin{bmatrix}
\mid & \mid & & \mid \\
T(\mathbf{e}_1) & T(\mathbf{e}_2) & \cdots & T(\mathbf{e}_n) \\
\mid & \mid & & \mid
\end{bmatrix}
$$


**Explanation:**

A matrix representation is obtained by applying the transformation to the standard basis vectors and placing the outputs as columns. This probing procedure turns an abstract map into a concrete array and leads naturally to special cases such as [projection](./09_projection_matrix_representation.ipynb), [reflection](./10_reflection_matrix_representation.ipynb), and [rotation](./11_rotation_matrix_representation.ipynb) matrices.

**Example:**

If

$$
T(x, y) =
\begin{bmatrix}
x \\
y \\
x + y
\end{bmatrix},
$$

then

$$
M_T =
\begin{bmatrix}
1 & 0 \\
0 & 1 \\
1 & 1
\end{bmatrix}.
$$


```python
import sympy as sp

basis_vector_1 = sp.Matrix([1, 0])
basis_vector_2 = sp.Matrix([0, 1])
output_vector_1 = sp.Matrix([1, 0, 1])
output_vector_2 = sp.Matrix([0, 1, 1])
representation_matrix = sp.Matrix.hstack(output_vector_1, output_vector_2)

print("representation_matrix =")
sp.pprint(representation_matrix)
```
