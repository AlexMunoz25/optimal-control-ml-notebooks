### 3.6.6. Combinatorial Optimization

$$
\min_{\pi\in\Pi}\; \sum_{i=1}^{n} C_{i,\pi(i)},
$$
where $\Pi$ is the set of permutations of $\{1,\dots,n\}$ (the assignment problem).

**Explanation:**

Combinatorial optimization selects the best configuration from a finite but astronomically large set of discrete structures — permutations, matchings, tours, subsets. The assignment problem shown here pairs $n$ agents with $n$ tasks to minimize total cost; the search space has $n!$ permutations. Many combinatorial problems admit compact [integer-programming](./01_integer_programming.ipynb) formulations, and structured ones like assignment are solvable in polynomial time by specialized algorithms rather than enumeration.

**Numerical Example:**

Cost matrix for three agents (rows) and three tasks (columns):
$$
C = \begin{bmatrix} 9 & 2 & 7 \\ 6 & 4 & 3 \\ 5 & 8 & 1 \end{bmatrix}.
$$

Enumerating the $3! = 6$ assignments, the permutation $\pi = (2,1,3)$ — agent 1 → task 2, agent 2 → task 1, agent 3 → task 3 — gives
$$
C_{1,2} + C_{2,1} + C_{3,3} = 2 + 6 + 1 = 9,
$$
which is the minimum total cost over all six permutations.

```python
import itertools

cost_matrix = [[9, 2, 7],
               [6, 4, 3],
               [5, 8, 1]]
agent_count = len(cost_matrix)

def assignment_cost(permutation):
    return sum(cost_matrix[agent][permutation[agent]] for agent in range(agent_count))

all_assignments = list(itertools.permutations(range(agent_count)))
best_assignment = min(all_assignments, key=assignment_cost)

print("all assignment costs:", {perm: assignment_cost(perm) for perm in all_assignments})
print("best assignment (agent -> task):", best_assignment, " cost:", assignment_cost(best_assignment))
```

**Equivalent scipy implementation:**

```python
import numpy as np
from scipy.optimize import linear_sum_assignment

cost_matrix = np.array([[9, 2, 7],
                        [6, 4, 3],
                        [5, 8, 1]])
agents, tasks = linear_sum_assignment(cost_matrix)

print("best assignment (agent -> task):", tuple(int(task) for task in tasks))
print("cost:", int(cost_matrix[agents, tasks].sum()))
```
