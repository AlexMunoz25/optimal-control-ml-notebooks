### 3.2.3.9. Conjugate Functions and Fenchel Inequality


$$f^*(y)=\sup_x(y^Tx-f(x)),\qquad f(x)+f^*(y)\geq y^Tx$$


**Explanation:**

The conjugate function records the largest affine gap with a chosen slope. For a strictly convex quadratic, the maximizing point is found by setting a derivative to zero. This construction underlies Fenchel inequalities and many dual functions.

The conjugate $f^*(y)$ takes the supremum over $x$ of a linear term $y^Tx$ minus $f(x)$. The inequality says the sum $f(x)+f^*(y)$ always dominates that [inner product](../../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/02_inner_product_space.ipynb) term.

**Properties:**
- The conjugate is a supremum over the primal variable.
- A derivative condition identifies the maximizer in smooth one-dimensional examples.
- Fenchel inequality follows directly from the supremum definition.

**Intuition:**

The figure above relates slopes to affine gaps; the arithmetic maximizes one such gap.

**Numerical Example:**

The figure above relates slopes to affine gaps; the arithmetic maximizes one such gap.

For $f(x)=\frac12x^2$, evaluate the conjugate at slope $y=3$:
$$
f^*(3)=\sup_x \left(3x-\frac12x^2\right).
$$

Differentiate the expression inside the supremum:
$$
\frac{d}{dx}\left(3x-\frac12x^2\right)=3-x.
$$

Set $3-x=0$, giving $x=3$. Substitute back:
$$
3\cdot3-\frac12\cdot3^2=9-\frac92=\frac92.
$$


```python
import sympy as sp

variable = sp.symbols('x')
expression = 3 * variable - sp.Rational(1, 2) * variable ** 2
optimizer = sp.solve(sp.diff(expression, variable), variable)[0]
value = sp.simplify(expression.subs(variable, optimizer))

print("optimizer =", optimizer)
print("value =", value)
```

**Direct `casadi` implementation:**

```python
import casadi as ca

variable = ca.SX.sym("x")
expression = 3 * variable - 0.5 * variable ** 2
gradient = ca.gradient(expression, variable)
hessian = ca.jacobian(gradient, variable)
optimizer = ca.solve(ca.evalf(hessian), -ca.evalf(ca.substitute(gradient, variable, ca.DM(0))))
value = ca.evalf(ca.substitute(expression, variable, optimizer))

print("optimizer =", optimizer)
print("value =", value)
```
