### 3.2.11.1. Inequality Constrained Convex Problems


$$\min_x f_0(x)\quad\text{subject to}\quad f_i(x)\leq0$$


**Explanation:**

Inequality-constrained convex problems combine a convex objective with convex inequality constraints and affine equalities. Interior-point methods work from strictly feasible points, so the distinction between boundary feasibility and strict interior feasibility matters immediately.

The variable $x$ is the decision vector, $f_0(x)$ is the objective, and each $f_i(x)\leq0$ is an inequality constraint. A feasible point must satisfy every inequality before its objective value is compared.

**Properties:**
- The log barrier requires every inequality value to be negative.
- The barrier is finite in the strict interior.
- Interior-point methods maintain strict feasibility while reducing the approximation gap.

**Intuition:**

The example evaluates one barrier contribution at a strictly feasible point.


**Numerical Example:**

For the constraint $f_1(x)=x-2<0$, evaluate the point $x=1$. Then
$$
f_1(1)=1-2=-1.
$$

The point is strictly feasible because $-1<0$. The log-barrier contribution is
$$
-\log(-f_1(1))=-\log(-(-1))=-\log(1)=0.
$$

The finite value confirms that the point is in the strict interior.


```python
import sympy as sp

point = sp.Integer(1)
constraint_value = point - 2
barrier_value = -sp.log(-constraint_value)

print("barrier_value =", barrier_value)
```
