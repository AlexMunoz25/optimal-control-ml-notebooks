### 3.4.11. Basis Pursuit and Sparse Descriptions

$$\min_x\|x\|_1\quad\text{subject to}\quad Ax=b$$


**Explanation:**

Basis pursuit seeks a sparse representation by minimizing the one-[norm](../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb) subject to exact linear equations. The one-norm is convex and often promotes sparse solutions without directly counting nonzero entries. Feasibility is exact representation; optimality is smallest one-norm among feasible representations.

The constraint $Ax=b$ requires an exact representation of the data vector $b$. The one-norm objective promotes a sparse coefficient vector $x$ while keeping that linear representation exact.

**Properties:**
- The equality constraint preserves the represented signal.
- The one-norm is a convex sparsity surrogate.
- Candidate feasible points can be compared by one-norm.

**Intuition:**

The example compares two feasible representations of the same scalar equation.


**Numerical Example:**

Consider
$$
x_1+2x_2=1.
$$

Two feasible points are
$$
x^{(1)}=\begin{bmatrix}1\\0\end{bmatrix},
\qquad
x^{(2)}=\begin{bmatrix}0\\1/2\end{bmatrix}.
$$

Check feasibility:
$$
1+2\cdot0=1,
\qquad
0+2\cdot\frac12=1.
$$

Compare one-norms:
$$
\|x^{(1)}\|_1=1,
\qquad
\|x^{(2)}\|_1=\frac12.
$$

The second feasible representation has the smaller one-norm.


```python
import sympy as sp

candidate_1 = sp.Matrix([1, 0])
candidate_2 = sp.Matrix([0, sp.Rational(1, 2)])
representation_values = [candidate_1[0] + 2 * candidate_1[1], candidate_2[0] + 2 * candidate_2[1]]
l1_norms = [sum(abs(entry) for entry in candidate) for candidate in [candidate_1, candidate_2]]

print("representation_values =", representation_values)
print("l1_norms =", l1_norms)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

candidate_1 = ca.DM([1, 0])
candidate_2 = ca.DM([0, 0.5])
representation_values = [float(candidate_1[0] + 2 * candidate_1[1]), float(candidate_2[0] + 2 * candidate_2[1])]
l1_norms = [float(ca.norm_1(candidate_1)), float(ca.norm_1(candidate_2))]

print("representation_values =", representation_values)
print("l1_norms =", l1_norms)
```
