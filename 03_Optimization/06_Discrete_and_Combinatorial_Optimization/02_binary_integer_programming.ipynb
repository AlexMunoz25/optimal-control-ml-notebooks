### 3.6.2. Binary Integer Programming (BIP)

$$
\begin{aligned}
\max_{\mathbf{x}}\quad & \mathbf{v}^\top \mathbf{x} \\
\text{subject to}\quad & \mathbf{w}^\top \mathbf{x} \le W,\\
& \mathbf{x}\in\{0,1\}^n.
\end{aligned}
$$

**Explanation:**

A binary integer program (BIP) restricts every variable to $\{0,1\}$, so each variable encodes a yes/no decision. This is the natural model for selection, assignment, and logical-condition problems; the 0–1 knapsack shown here selects a subset of items of maximum value subject to a capacity limit. With $n$ variables the feasible set has at most $2^n$ points, so complete enumeration is possible for small $n$ but grows exponentially, motivating the bounding techniques of [integer programming](./01_integer_programming.ipynb).

**Numerical Example:**

Four items with values $\mathbf{v} = (3,4,5,6)$ and weights $\mathbf{w} = (2,3,4,5)$, capacity $W = 5$:
$$
\max\; 3x_1 + 4x_2 + 5x_3 + 6x_4
\quad\text{s.t.}\quad 2x_1 + 3x_2 + 4x_3 + 5x_4 \le 5,\; x_i\in\{0,1\}.
$$

Checking subsets whose weight is at most $5$: $\{1,2\}$ has weight $5$ and value $7$; the single item $\{4\}$ has value $6$; $\{3\}$ has value $5$. The best is
$$
\mathbf{x}^\star = (1,1,0,0),\qquad \text{value } 7,\ \text{weight } 5.
$$

```python
import itertools

values = [3, 4, 5, 6]
weights = [2, 3, 4, 5]
capacity = 5
item_count = len(values)

def total(selection, table):
    return sum(table[i] for i in range(item_count) if selection[i])

feasible_selections = [selection
                       for selection in itertools.product([0, 1], repeat=item_count)
                       if total(selection, weights) <= capacity]
best_selection = max(feasible_selections, key=lambda selection: total(selection, values))

print("best selection:", best_selection)
print("value:", total(best_selection, values), " weight:", total(best_selection, weights))
```

**Equivalent scipy.milp implementation:**

```python
import numpy as np
from scipy.optimize import milp, LinearConstraint, Bounds

cost = -np.array([3.0, 4.0, 5.0, 6.0])
capacity_constraint = LinearConstraint(np.array([[2.0, 3.0, 4.0, 5.0]]), -np.inf, 5)
result = milp(c=cost, constraints=capacity_constraint, integrality=np.ones(4),
              bounds=Bounds(lb=0, ub=1))

print("best selection:", result.x.round().astype(int), " value:", -result.fun)
```
