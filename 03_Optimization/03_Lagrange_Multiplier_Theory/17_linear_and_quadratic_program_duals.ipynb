### 3.3.17. Linear and Quadratic Program Duals

$$
\min_{Ex=d,\ x\ge 0} c^T x
\quad \Longleftrightarrow \quad
\max_{E^T\lambda\le c} d^T\lambda
$$

$$
q(\mu)=\inf_x\left(\frac{1}{2}x^TQx+c^Tx+\mu^T(Ax-b)\right)
$$

**Explanation:**

Linear programming duality follows by minimizing a linear Lagrangian over the nonnegative orthant: finite dual values occur exactly when the Lagrangian coefficients are nonnegative. Quadratic programming duality follows by minimizing a strictly convex quadratic Lagrangian explicitly, using the [matrix inverse](../../01_Foundations/01_Linear_Algebra/03_Matrix/11_matrix_inverse.ipynb) of the positive definite matrix $Q$. Both cases show how multiplier variables become the decision variables of a related optimization problem.

**Numerical Example:**

Consider the linear program

$$
\min x_1+2x_2
\quad \text{subject to} \quad x_1+x_2=1,
\qquad x_1,x_2\ge 0.
$$

Here

$$
E=\begin{bmatrix}1&1\end{bmatrix},
\qquad d=1,
\qquad c=\begin{bmatrix}1\\2\end{bmatrix}.
$$

The dual is

$$
\max \lambda
\quad \text{subject to} \quad
\lambda\le 1,
\qquad \lambda\le 2.
$$

The largest feasible value is

$$
\lambda^*=1.
$$

The dual value is

$$
d^T\lambda^*=1\cdot 1=1.
$$

The primal feasible point $x^*=(1,0)$ has cost

$$
c^T x^*=1\cdot 1+2\cdot 0=1.
$$

Complementary slackness also matches:

$$
c-E^T\lambda^*=\begin{bmatrix}1\\2\end{bmatrix}-\begin{bmatrix}1\\1\end{bmatrix}
=\begin{bmatrix}0\\1\end{bmatrix},
$$

so the positive primal coordinate $x_1^*=1$ corresponds to zero dual slack, while $x_2^*=0$ may have positive dual slack.

```python
import sympy as sp

E = sp.Matrix([[1, 1]])
cost_vector = sp.Matrix([1, 2])
primal_solution = sp.Matrix([1, 0])
dual_multiplier = sp.Integer(1)

dual_slack = cost_vector - E.T * dual_multiplier
primal_value = (cost_vector.T * primal_solution)[0]
dual_value = dual_multiplier
complementarity_products = sp.Matrix([
    primal_solution[0] * dual_slack[0],
    primal_solution[1] * dual_slack[1],
])

result = {
    'dual_slack': dual_slack,
    'primal_value': primal_value,
    'dual_value': dual_value,
    'complementarity_products': complementarity_products,
}

for label, value in result.items():
    print(f"{label} =", value)
```
