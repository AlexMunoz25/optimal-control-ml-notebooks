### 1.1.8.9. Einstein Summation Convention


$$
A'^{\,i} = a^{i}_{\;j} A^j
$$


**Explanation:**

Einstein summation convention suppresses explicit summation signs when one index is repeated once up and once down. This notation compresses the transformation laws for covariant and contravariant components and pairs naturally with the [Jacobian matrix](./10_jacobian_matrix.ipynb), which supplies the actual coefficients.

**Example:**

The repeated index $j$ is a dummy index and is summed over, while the free index $i$ labels the output component.


```python
import numpy as np

matrix_a = np.array([[0.0, 1.0], [-1.0, 0.0]])
vector_a = np.array([2.0, 3.0])
transformed_vector = matrix_a @ vector_a

print("transformed_vector =", transformed_vector)
```
