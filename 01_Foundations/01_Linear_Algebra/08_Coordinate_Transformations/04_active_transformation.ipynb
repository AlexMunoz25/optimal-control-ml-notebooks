### 1.1.8.4. Active Transformation


$$
\begin{aligned}
\mathbf{A}' &= R_{\mathrm{active}}(\alpha)\,\mathbf{A}, \\
R_{\mathrm{active}}(\alpha)
&=
\begin{bmatrix}
\cos \alpha & -\sin \alpha \\
\sin \alpha & \cos \alpha
\end{bmatrix}, \\
A_x' &= A_x \cos \alpha - A_y \sin \alpha,
\qquad
A_y' = A_x \sin \alpha + A_y \cos \alpha
\end{aligned}
$$


**Explanation:**

An active transformation rotates the vector itself while the coordinate axes stay fixed. The same direct matrix used to rotate basis vectors also rotates any vector expressed in that basis. This differs from a [passive transformation](./03_passive_transformation.ipynb), where the vector is unchanged and only its coordinates are updated.

</p>

**Example:**

For
$$
\mathbf{A} =
\begin{bmatrix}
5 \\
3
\end{bmatrix}
$$
and a $150^{\circ}$ active rotation,
$$
\mathbf{A}'
=
\begin{bmatrix}
\cos 150^{\circ} & -\sin 150^{\circ} \\
\sin 150^{\circ} & \cos 150^{\circ}
\end{bmatrix}
\begin{bmatrix}
5 \\
3
\end{bmatrix}
=
\begin{bmatrix}
-5.83 \\
-0.10
\end{bmatrix}.
$$


```python
import numpy as np

rotation_angle_degrees = 150.0
rotation_angle = np.deg2rad(rotation_angle_degrees)
active_transformation_matrix = np.array(
    [
        [np.cos(rotation_angle), -np.sin(rotation_angle)],
        [np.sin(rotation_angle), np.cos(rotation_angle)],
    ]
)
original_vector_components = np.array([5.0, 3.0])
rotated_vector_components = active_transformation_matrix @ original_vector_components

print("active_transformation_matrix =")
print(np.round(active_transformation_matrix, 4))
print("rotated_vector_components =", np.round(rotated_vector_components, 4))
```
