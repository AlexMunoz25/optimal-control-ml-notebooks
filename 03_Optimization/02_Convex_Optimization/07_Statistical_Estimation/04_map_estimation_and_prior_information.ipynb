### 3.2.7.4. MAP Estimation and Prior Information


$$\max_\theta\log p(y\mid\theta)+\log p(\theta)$$


**Explanation:**

Maximum a posteriori estimation adds prior information to the likelihood objective. In convex form, the negative log-prior becomes a regularizer added to the data-fit term. Feasibility is the parameter domain, and optimality balances evidence from data against the prior penalty.

The first log term measures how well parameter $\theta$ explains the data $y$. The second log term adds prior information about $\theta$, so MAP estimation maximizes likelihood plus prior log-density.

**Properties:**
- Quadratic priors produce quadratic regularization.
- MAP and regularized least squares have the same algebra in Gaussian scalar examples.
- The prior shifts the estimate toward its preferred value.

**Intuition:**

The scalar example shows the prior pulling the optimizer away from the unregularized data target.


**Numerical Example:**

Minimize the scalar MAP objective
$$
f(x)=(x-2)^2+3x^2.
$$

Expand:
$$
f(x)=x^2-4x+4+3x^2=4x^2-4x+4.
$$

Differentiate:
$$
f'(x)=8x-4.
$$

Set the derivative to zero:
$$
8x-4=0\quad\Longrightarrow\quad x^\star=\frac12.
$$

The prior penalty pulls the estimate toward zero.


```python
import sympy as sp

variable = sp.symbols('x')
objective = (variable - 2) ** 2 + 3 * variable ** 2
optimizer = sp.solve(sp.diff(objective, variable), variable)[0]

print("optimizer =", optimizer)
```

**Direct `casadi` implementation:**

```python
import casadi as ca

variable = ca.SX.sym("x")
objective = (variable - 2) ** 2 + 3 * variable ** 2
gradient = ca.gradient(objective, variable)
hessian = ca.jacobian(gradient, variable)
optimizer = ca.solve(ca.evalf(hessian), -ca.evalf(ca.substitute(gradient, variable, ca.DM(0))))

print("optimizer =", optimizer)
```
