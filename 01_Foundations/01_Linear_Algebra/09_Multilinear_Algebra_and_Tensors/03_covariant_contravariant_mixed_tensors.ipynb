### 1.1.9.3. Covariant, Contravariant, and Mixed Tensors

$$
\begin{aligned}
A'^{\,ij} &= \frac{\partial x'^{\,i}}{\partial x^k}\,\frac{\partial x'^{\,j}}{\partial x^l}\,A^{kl}, \\
A'_{ij} &= \frac{\partial x^k}{\partial x'^{\,i}}\,\frac{\partial x^l}{\partial x'^{\,j}}\,A_{kl}, \\
A'^{\,i}{}_j &= \frac{\partial x'^{\,i}}{\partial x^k}\,\frac{\partial x^l}{\partial x'^{\,j}}\,A^k{}_l
\end{aligned}
$$

**Explanation:**

A second-rank tensor can carry two upper indices, two lower indices, or one of each. A super-index such as the $i$ in $A^{ij}$ is an upper index, so it marks a contravariant slot; a sub-index such as the $j$ in $A_{ij}$ is a lower index, so it marks a covariant slot. Each upper index transforms like the components from [Contravariant Components](../08_Coordinate_Transformations/07_contravariant_components.ipynb), and each lower index transforms like the components from [Covariant Components](../08_Coordinate_Transformations/08_covariant_components.ipynb). Mixed tensors therefore apply the direct Jacobian in one slot and the inverse Jacobian in another.

**Example:**

For the coordinate change
$$
x'^1 = 2x^1,
\qquad
x'^2 = \frac{1}{2} x^2,
\qquad
J =
\begin{bmatrix}
2 & 0 \\
0 & \frac{1}{2}
\end{bmatrix},
$$
a contravariant tensor
$$
A^{ij} =
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}
$$
transforms to
$$
A'^{\,ij} = J A J^\top =
\begin{bmatrix}
4 & 2 \\
3 & 1
\end{bmatrix}.
$$

```python
import sympy as sp

coordinate_jacobian = sp.Matrix([[2, 0], [0, sp.Rational(1, 2)]])
inverse_coordinate_jacobian = coordinate_jacobian.inv()
tensor_components = sp.Matrix([[1, 2], [3, 4]])

contravariant_tensor_components = coordinate_jacobian * tensor_components * coordinate_jacobian.T
covariant_tensor_components = (
    inverse_coordinate_jacobian.T * tensor_components * inverse_coordinate_jacobian
)
mixed_tensor_components = coordinate_jacobian * tensor_components * inverse_coordinate_jacobian

print("contravariant_tensor_components =")
sp.pprint(contravariant_tensor_components)
print("covariant_tensor_components =")
sp.pprint(covariant_tensor_components)
print("mixed_tensor_components =")
sp.pprint(mixed_tensor_components)
```
