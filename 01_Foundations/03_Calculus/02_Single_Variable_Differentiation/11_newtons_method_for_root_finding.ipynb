### 1.3.2.11. Newton's Method for Root Finding

$$
x_{n} = x_{n-1} - \frac{f(x_{n-1})}{f'(x_{n-1})} .
$$

**Explanation:**

Newton's method solves $f(x) = 0$ by repeatedly replacing $f$ with its [linear approximation](./10_linear_approximation.ipynb) and jumping to where that tangent line crosses zero. Each step uses the value and slope at the current guess, and near a simple root the convergence is quadratic — the number of correct digits roughly doubles per iteration. This is the root-finding primitive; applying the same idea to $f' = 0$ gives [Newton's method for minimization](../../../03_Optimization/05_Unconstrained_Algorithms/11_newtons_method.ipynb). It is fast but local: a poor start, or $f'(x_n) \approx 0$, can make it diverge, which is why it is often paired with the bracketing guarantee of the [intermediate value theorem](../01_Limits_and_Continuity/06_intermediate_value_theorem.ipynb).

**Assumptions:**
- $f$ is differentiable and $f'(x_n) \neq 0$ at each iterate.
- The initial guess $x_0$ is sufficiently close to the root.

**Intuition:**

Each iterate is the $x$-intercept of the tangent line at the previous guess, so the approximations $x_0, x_1, x_2, \ldots$ march toward the root $x^\star$.

**Numerical Example:**

Approximate a root of $f(x) = x^3 - 3x + 1$ with $x_0 = 2$. The derivative is $f'(x) = 3x^2 - 3$, so

$$
x_1 = 2 - \frac{f(2)}{f'(2)} = 2 - \frac{3}{9} \approx 1.6667,
$$

$$
x_2 = x_1 - \frac{f(x_1)}{f'(x_1)} \approx 1.5486, \qquad x_3 \approx 1.5322, \qquad x_4 \approx 1.5321 .
$$

The iterates stabilize at $x \approx 1.53209$, a root of the cubic; the rapid agreement of $x_3$ and $x_4$ reflects the quadratic convergence.

```python
import sympy as sp

x = sp.symbols("x")
f = x**3 - 3 * x + 1
derivative = sp.diff(f, x)
newton_update = x - f / derivative

iterate = sp.Rational(2)
print("x_0 =", float(iterate))
for step in range(1, 5):
    iterate = newton_update.subs(x, iterate)
    print(f"x_{step} =", sp.N(iterate, 10))
```

**Equivalent CasADi implementation:**

CasADi's Newton root-finder applies the same update internally, converging to the same root from $x_0 = 2$.

```python
import casadi as ca

x = ca.SX.sym("x")
residual = x**3 - 3 * x + 1
root_finder = ca.rootfinder("newton_root", "newton", {"x": x, "g": residual})

solution = root_finder(x0=2.0)
print("CasADi Newton root =", float(solution["x"]))
```
