### 1.1.2.6. Norm of a Vector (Euclidean)

$$
\|\mathbf{v}\| = \sqrt{\sum_{i=1}^{n} v_i^2} = \sqrt{v_1^2 + v_2^2 + \cdots + v_n^2}
$$

**Explanation:**

The Euclidean norm measures the magnitude (length) of a vector.
It is useful in optimization and distance computations in ML.

For a [vector](./01_vector_definition.ipynb), this magnitude also appears in the geometric formula for the [dot product](./04_vector_dot_product.ipynb) and later extends to the [matrix Frobenius norm](../03_Matrix/15_matrix_frobenius_norm.ipynb).

**Example:**

If

$$
\mathbf{v} =
\begin{bmatrix}
3 \\
4
\end{bmatrix},
$$

then

$$
\|\mathbf{v}\| = \sqrt{3^2 + 4^2} = 5.
$$


```python
import sympy as sp

vector_v = sp.Matrix([3, 4])

euclidean_norm = vector_v.norm()

print("‖v‖ =", euclidean_norm)
```
