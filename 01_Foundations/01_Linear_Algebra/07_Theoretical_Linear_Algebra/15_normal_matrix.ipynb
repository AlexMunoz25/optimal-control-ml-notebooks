### 1.1.7.15. Normal Matrix


$$
A^{\mathsf{T}} A = A A^{\mathsf{T}}
$$


**Explanation:**

A normal matrix commutes with its transpose. This structural condition guarantees an orthogonal eigendecomposition and explains why important subclasses such as [symmetric matrices](./09_symmetric_matrix.ipynb) and [orthogonal matrices](./11_orthogonal_matrix.ipynb) fit naturally into the same picture.

**Example:**

A symmetric matrix is automatically normal, so it can be diagonalized with an orthogonal change-of-basis matrix.


```python
import sympy as sp

matrix_a = sp.Matrix([[2, 1], [1, 3]])

print("left_side =")
sp.pprint(matrix_a.T * matrix_a)
print("right_side =")
sp.pprint(matrix_a * matrix_a.T)
print("is_normal =", matrix_a.T * matrix_a == matrix_a * matrix_a.T)
```
