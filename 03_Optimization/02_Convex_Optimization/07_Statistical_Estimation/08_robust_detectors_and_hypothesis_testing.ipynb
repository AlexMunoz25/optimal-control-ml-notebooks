### 3.2.7.8. Robust Detectors and Hypothesis Testing


$$\min\max_{p\in\mathcal P}\operatorname{risk}(p)$$


**Explanation:**

Robust detector design optimizes performance over a family of possible distributions or risk models. The worst-case risk is a maximum over scenarios, and maxima of affine or convex risk functions preserve convexity. Feasibility is the detector class; optimality minimizes the largest risk.

The outer minimization chooses a detector or decision rule. The inner maximum ranges over distributions $p$ in the uncertainty set $\mathcal P$, so the objective is worst-case risk.

**Properties:**
- Robust objectives use a maximum over uncertainty cases.
- The maximum of convex functions is convex.
- The active case is the one with largest current risk.

**Intuition:**

The example computes the worst-case risk from two possible models.


**Numerical Example:**

Suppose two admissible models produce risks
$$
r_1=1/5,\qquad r_2=3/10.
$$

The robust risk is
$$
\max\{r_1,r_2\}=\max\{1/5,3/10\}.
$$

Since
$$
1/5=2/10<3/10,
$$
the worst-case risk is
$$
3/10.
$$


```python
import sympy as sp

risks = [sp.Rational(1, 5), sp.Rational(3, 10)]
worst_risk = max(risks)

worst_risk
```
