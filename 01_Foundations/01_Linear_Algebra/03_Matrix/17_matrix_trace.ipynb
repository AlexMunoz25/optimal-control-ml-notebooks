### 1.1.3.17. Trace of a Matrix

$$
\text{tr}(\mathbf{A}) = \sum_{i=1}^n A_{ii}
$$

**Explanation:**

The trace of a square matrix $\mathbf{A} \in \mathbb{R}^{n \times n}$ is the sum of its diagonal entries.

**Properties:**
- Linear: $\text{tr}(\mathbf{A} + \mathbf{B}) = \text{tr}(\mathbf{A}) + \text{tr}(\mathbf{B})$
- Invariant under cyclic permutations: $\text{tr}(\mathbf{AB}) = \text{tr}(\mathbf{BA})$
- Equal to the sum of eigenvalues (counted with algebraic multiplicity)

**Related concepts:**
- Builds on [Matrices](./01_matrix_definition.ipynb).
- The cyclic property is used more explicitly in [Trace of a Matrix Product](./18_matrix_trace_of_product.ipynb) and also appears in [Matrix Frobenius Norm](./15_matrix_frobenius_norm.ipynb).

**Example:**

For

$$
\mathbf{A} =
\begin{bmatrix}
2 & 1 & 0 \\
0 & 3 & 4 \\
0 & 0 & 5
\end{bmatrix},
$$

the trace is

$$
\text{tr}(\mathbf{A}) = 2 + 3 + 5 = 10.
$$


```python
import sympy as sp

matrix_a = sp.Matrix([[1, 2], [3, 4]])

trace_value = matrix_a.trace()

print("Tr(A) =", trace_value)
```
