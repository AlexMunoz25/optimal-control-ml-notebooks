### 1.3.1.7. Extreme Value Theorem

$$
f \text{ continuous on the closed, bounded interval } [a,b]
\;\;\Longrightarrow\;\;
\exists\, c, d \in [a,b] : f(c) \le f(x) \le f(d)\ \ \forall x \in [a,b] .
$$

**Explanation:**

If $f$ is continuous on a closed, bounded interval $[a,b]$, then $f$ attains an absolute maximum and an absolute minimum on that interval — the values $f(c)$ and $f(d)$ are actually achieved, not merely approached. Both hypotheses are essential: on an open interval, or with a single discontinuity, an extremum can fail to exist. This is the existence result the optimization track relies on when it asserts that a minimizer of a [cost function](../../../03_Optimization/01_Unconstrained_Optimization/05_quadratic_cost_functions_and_existence.ipynb) exists before any algorithm searches for it. The points where extrema occur are the endpoints together with the interior [critical points](../02_Single_Variable_Differentiation/08_monotonicity_and_first_derivative_test.ipynb).

**Numerical Example:**

Take $f(x) = x^2$ on the closed, bounded interval $[-1, 2]$. The theorem guarantees both extrema exist; we locate them among the endpoints and interior critical points. The derivative is

$$
f'(x) = 2x = 0 \;\Rightarrow\; x = 0 \in (-1, 2),
$$

the only critical point. Evaluating the candidates,

$$
f(-1) = 1, \qquad f(0) = 0, \qquad f(2) = 4 .
$$

Therefore the absolute minimum is $f(0) = 0$ (attained at the interior critical point) and the absolute maximum is $f(2) = 4$ (attained at an endpoint).

By contrast, $g(x) = x$ on the *open* interval $(0,1)$ attains neither bound — its values fill $(0,1)$ without reaching $0$ or $1$ — showing the closed-interval hypothesis cannot be dropped.

```python
import sympy as sp

x = sp.symbols("x")
f = x**2
left_end, right_end = sp.Integer(-1), sp.Integer(2)

derivative = sp.diff(f, x)
interior_critical = [point for point in sp.solve(derivative, x) if left_end < point < right_end]
candidate_points = [left_end, right_end] + interior_critical
candidate_values = {point: f.subs(x, point) for point in candidate_points}

minimizer = min(candidate_values, key=lambda point: candidate_values[point])
maximizer = max(candidate_values, key=lambda point: candidate_values[point])

print("f'(x) =", derivative, " -> interior critical points:", interior_critical)
for point in candidate_points:
    print(f"f({point}) = {candidate_values[point]}")
print("absolute minimum:", candidate_values[minimizer], "at x =", minimizer)
print("absolute maximum:", candidate_values[maximizer], "at x =", maximizer)
```
