### 1.1.9.5. Tensor Outer Product

$$
C^{ij} = A^i B^j,
\qquad
C^{ik}{}_{jlm} = A^i{}_j B^k{}_{lm},
\qquad
\operatorname{rank}(C) = \operatorname{rank}(A) + \operatorname{rank}(B)
$$

**Explanation:**

The outer product multiplies tensor components without summing over any index. Every index from the input tensors remains in the result, so the ranks add. This extends [Matrix Outer Product](../03_Matrix/14_matrix_outer_product.ipynb) from vectors to arbitrary tensors.

**Example:**

For

$$
\mathbf{a}=\begin{bmatrix}1\\2\\3\end{bmatrix},
\qquad
\mathbf{b}=\begin{bmatrix}4\\5\\6\end{bmatrix},
$$

the outer product has entries $a_i b_j$:

$$
\mathbf{a}\otimes\mathbf{b}
=\begin{bmatrix}
1\cdot4&1\cdot5&1\cdot6\\
2\cdot4&2\cdot5&2\cdot6\\
3\cdot4&3\cdot5&3\cdot6
\end{bmatrix}
=\begin{bmatrix}4&5&6\\8&10&12\\12&15&18\end{bmatrix}.
$$

The outer product of two $2\times2$ tensors has shape $(2,2,2,2)$.


```python
import sympy as sp

vector_a = sp.Array([1, 2, 3])
vector_b = sp.Array([4, 5, 6])
rank_2_outer_product = sp.tensorproduct(vector_a, vector_b)

matrix_factor = sp.Array([[1, 0], [0, 2]])
rank_4_outer_product = sp.tensorproduct(matrix_factor, matrix_factor)

print("rank_2_outer_product =")
sp.pprint(rank_2_outer_product)
print("rank_4_outer_product shape =", rank_4_outer_product.shape)
```
