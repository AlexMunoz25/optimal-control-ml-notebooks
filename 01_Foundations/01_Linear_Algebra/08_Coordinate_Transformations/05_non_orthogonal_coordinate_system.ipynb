### 1.1.8.5. Non-Orthogonal Coordinate System


$$
\mathbf{A} = A^1 \mathbf{e}_1 + A^2 \mathbf{e}_2
$$


**Explanation:**

Fleisch introduces non-orthogonal coordinates to show that projection is no longer unique once the axes are oblique. The core concept is the coordinate system itself: the angle between basis directions matters, so parallel and perpendicular projections become different operations. The later notebooks on [contravariant](./07_contravariant_components.ipynb) and [covariant components](./08_covariant_components.ipynb) explain those two projection choices separately.

**Example:**

In a skew basis with a $60^{\circ}$ angle between axes, the same vector can be described by coefficients that no longer coincide with perpendicular shadows on the axes.


```python
import numpy as np

basis_angle = np.deg2rad(60.0)
basis_e1 = np.array([1.0, 0.0])
basis_e2 = np.array([np.cos(basis_angle), np.sin(basis_angle)])

print("basis_e1 =", np.round(basis_e1, 4))
print("basis_e2 =", np.round(basis_e2, 4))
```
