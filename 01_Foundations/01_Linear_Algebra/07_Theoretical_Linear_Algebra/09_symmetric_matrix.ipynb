### 1.1.7.9. Symmetric Matrix


$$
A^{\mathsf{T}} = A
$$


**Explanation:**

A symmetric matrix equals its transpose. Savov highlights real eigenvalues and orthogonal eigenvectors as the main structural consequences, so this notebook concentrates on those properties rather than on positive definiteness, which is treated separately in [positive definite matrices](./10_positive_definite_matrix.ipynb).

**Example:**

The matrix

$$
\begin{bmatrix}
2 & 1 \\
1 & 3
\end{bmatrix}
$$

is symmetric because the off-diagonal entries match.


```python
import sympy as sp

matrix_a = sp.Matrix([[2, 1], [1, 3]])

print("is_symmetric =", matrix_a == matrix_a.T)
print("eigenvalues =", list(matrix_a.eigenvals().keys()))
```

**References:**
