### 3.4.19. Sensitivity, Shadow Prices, and the Primal Function

$$
p(u)=f(x(u)), \qquad h(x(u))=u
$$

$$
\nabla p(u)=-\lambda(u)
$$

**Explanation:**

The primal function records the locally optimal cost after the equality constraint level is changed from $0$ to $u$. The sensitivity equation says that the multiplier is the negative derivative of this optimal value function. In words, $\lambda^*$ measures the first-order decrease in optimal cost produced by relaxing the constraint level.

**Intuition:**

<img src="../../Figures/030419_berts_fig_3_2_3_sensitivity_linear_constraint.jpeg" alt="Bertsekas Figure 3.2.3: sensitivity for a perturbed linear equality constraint" style="width: 60%; height: auto;">

<img src="../../Figures/030419_berts_fig_3_2_4_primal_function_sensitivity.jpeg" alt="Bertsekas Figure 3.2.4: primal function slope and Lagrange multiplier" style="width: 60%; height: auto;">

The figures connect a shifted constraint to a shifted local minimum. The slope of the resulting optimal value curve at the original constraint level is the negative multiplier.

**Numerical Example:**

Use the primal function example from the figure:

$$
f(x)=\frac{1}{2}(x_1^2-x_2^2)-x_2,
\qquad h(x)=x_2.
$$

Perturb the constraint to

$$
h(x)=u \quad \Rightarrow \quad x_2=u.
$$

With $x_2=u$, the constrained cost becomes

$$
f(x_1,u)=\frac{1}{2}(x_1^2-u^2)-u.
$$

Minimizing over $x_1$ gives

$$
\frac{\partial f}{\partial x_1}=x_1=0,
$$

so

$$
x(u)=(0,u).
$$

The primal function is therefore

$$
p(u)=f(0,u)=-\frac{1}{2}u^2-u.
$$

Differentiate:

$$
p'(u)=-u-1,
\qquad
p'(0)=-1.
$$

The original multiplier was $\lambda^*=1$, so

$$
p'(0)=-\lambda^*.
$$

This matches the sensitivity theorem exactly.

```python
import sympy as sp

x_1, u = sp.symbols('x_1 u')

cost_under_perturbation = sp.Rational(1, 2) * (x_1**2 - u**2) - u
stationary_coordinate = sp.solve(sp.diff(cost_under_perturbation, x_1), x_1)[0]
primal_function = sp.simplify(cost_under_perturbation.subs(x_1, stationary_coordinate))
primal_derivative = sp.diff(primal_function, u)
lambda_at_zero = -primal_derivative.subs(u, 0)

result = {
    'x_1_of_u': stationary_coordinate,
    'primal_function': primal_function,
    'primal_derivative': primal_derivative,
    'lambda_at_zero': lambda_at_zero,
}

for label, value in result.items():
    print(f"{label} =", value)
```

**Equivalent `casadi` implementation:**

```python
import casadi as ca

x_1 = ca.SX.sym("x_1")
u = ca.SX.sym("u")
cost = 0.5 * (x_1 ** 2 - u ** 2) - u
gradient = ca.gradient(cost, x_1)
hessian = ca.jacobian(gradient, x_1)
stationary_coordinate = -ca.substitute(gradient, x_1, 0) / hessian
primal_function = ca.substitute(cost, x_1, stationary_coordinate)
primal_derivative = ca.gradient(primal_function, u)
lambda_at_zero = ca.evalf(-ca.substitute(primal_derivative, u, 0))

print("stationary_coordinate =", stationary_coordinate)
print("primal_derivative =", primal_derivative)
print("lambda_at_zero =", lambda_at_zero)
```

**Global perturbation and shadow prices:**

The sensitivity identity $\nabla p(u)=-\lambda(u)$ is local. For convex problems there is a matching *global* bound: perturbing the constraint right-hand sides by $u$ moves the optimal value according to

$$
p^*(u)\;\ge\;p^*(0)-\lambda^{*T}u,
$$

so the optimal multipliers $\lambda^*$ are supporting slopes of the value function. Read economically, each $\lambda_i^*$ is a **shadow price**: the marginal change in optimal cost per unit relaxation of constraint $i$. A large multiplier flags a constraint that is expensive to tighten, which is exactly the information used to price resources in a dual solution.
