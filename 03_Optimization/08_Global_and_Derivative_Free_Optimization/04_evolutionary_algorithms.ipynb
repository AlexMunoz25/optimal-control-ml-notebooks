### 3.8.4. Evolutionary Algorithms

$$
\text{population}_{k+1} = \text{select}\circ\text{mutate}\circ\text{recombine}\,(\text{population}_k),
$$
minimizing $f$ by survival of the fittest candidates.

**Explanation:**

Evolutionary algorithms maintain a *population* of candidate solutions and improve it over generations by selecting the fittest members, recombining them, and adding random mutation. They use only objective-function values — no gradients — so they apply to nonconvex, noisy, or discontinuous landscapes where [derivative-based methods](../09_Algorithms/01_descent_directions.ipynb) struggle, at the cost of many function evaluations. The $(\mu,\lambda)$ evolution strategy shown here keeps the best $\mu$ of $\lambda$ mutated offspring each generation.

**Assumptions:**
- The objective can be evaluated cheaply and repeatedly.
- No gradient information is available or reliable.

**Numerical Example:**

Minimize the multimodal $f(x) = x^2 + 10\big(1 - \cos(2\pi x)\big)$ (a 1-D Rastrigin function) on $[-5,5]$, whose global minimizer is $x=0$ with $f=0$ but which has many local minima at every integer. Starting from a random population and repeatedly keeping the best half while mutating with shrinking Gaussian steps, the best fitness falls toward $0$ within a few dozen generations — the population escapes the integer-spaced local minima that trap a single descent step.

```python
import numpy as np

def rastrigin(point):
    return point**2 + 10 * (1 - np.cos(2 * np.pi * point))

rng = np.random.default_rng(0)
population = rng.uniform(-5, 5, size=20)
step_size = 2.0
for generation in range(40):
    offspring = np.concatenate([population, population + rng.normal(0, step_size, size=population.size)])
    offspring = np.clip(offspring, -5, 5)
    survivors = offspring[np.argsort(rastrigin(offspring))[:20]]
    population = survivors
    step_size *= 0.92

best = population[int(np.argmin(rastrigin(population)))]
print("best solution x =", round(float(best), 4))
print("best fitness f =", round(float(rastrigin(best)), 4))
```
