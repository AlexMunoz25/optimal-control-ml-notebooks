### 1.4.1.3. Differential Equations as Mathematical Models

$$
\frac{dP}{dt} = kP, \qquad
\frac{dT}{dt} = k\,(T - T_m), \qquad
m\,\frac{dv}{dt} = mg - kv .
$$

**Explanation:**

A **mathematical model** is a description of a system in mathematical terms; when the description involves a *rate of change*, that rate becomes a [derivative](../../03_Calculus/02_Single_Variable_Differentiation/01_derivative_definition.ipynb) and the model becomes a differential equation. The pattern is always the same: state an assumption about how fast something changes, write it as a DE, solve, and check the prediction against reality. The three equations above model population growth (rate proportional to current size), Newton's law of cooling (rate proportional to the temperature gap), and a falling body with linear drag (Newton's second law) — the seeds of the [first-order](../03_Modeling_with_First_Order_Equations/01_growth_and_decay.ipynb) and [higher-order](../05_Modeling_with_Higher_Order_Equations/01_free_spring_mass_systems.ipynb) modeling studied later and the dynamics the control track builds on.

**Numerical Example:**

Model a population whose growth rate is proportional to its current size $P(t)$, with initial size $P_0$. The assumption "rate of change $\propto$ current amount" is written

$$
\frac{dP}{dt} = kP, \qquad P(0) = P_0 .
$$

Separating variables and integrating,

$$
\frac{dP}{P} = k\,dt \;\Rightarrow\; \ln P = kt + C_1 \;\Rightarrow\; P(t) = C e^{kt}.
$$

Applying the initial condition $P(0) = P_0$ gives $C = P_0$, so

$$
P(t) = P_0\,e^{kt}.
$$

The model predicts exponential growth when $k > 0$ and exponential decay when $k < 0$ — the same equation governs radioactive decay and continuously compounded interest.

```python
import sympy as sp

t, k = sp.symbols("t k")
P_0 = sp.symbols("P_0", positive=True)
P = sp.Function("P")

growth_model = sp.Eq(P(t).diff(t), k * P(t))
general_solution = sp.dsolve(growth_model, P(t))
particular_solution = sp.dsolve(growth_model, P(t), ics={P(0): P_0})

print("model            : dP/dt = kP")
print("general solution :", general_solution)
print("with P(0) = P_0  :", particular_solution)
```
