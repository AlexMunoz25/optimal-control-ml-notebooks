### 1.3.3.2. Definite Integral and Riemann Sums

$$
\int_{a}^{b} f(x)\,dx
= \lim_{n \to \infty} \sum_{i=1}^{n} f(x_i^{*})\,\Delta x,
\qquad \Delta x = \frac{b - a}{n} .
$$

**Explanation:**

The definite integral is the [limit](../01_Limits_and_Continuity/01_limit_of_a_function.ipynb) of Riemann sums: partition $[a,b]$ into $n$ subintervals of width $\Delta x$, build a rectangle of height $f(x_i^*)$ on each, add their signed areas, and let $n \to \infty$. The result is a single number — the net signed area between the graph and the $x$-axis. For a continuous function the choice of sample point $x_i^*$ (left, right, or midpoint) does not affect the limit. This "limit of a sum" is the template for every accumulated quantity downstream: a [cost functional](../../../04_Optimal_Control), an expectation, or a discounted return.

**Intuition:**

Rectangles on each subinterval approximate the area; refining the partition drives the approximation to the exact integral.

**Numerical Example:**

Estimate $\int_{0}^{2} x^2\,dx$ with a right-endpoint sum of $n = 4$ rectangles. Then $\Delta x = \tfrac{2-0}{4} = 0.5$ and the right endpoints are $x_i = 0.5, 1.0, 1.5, 2.0$:

$$
R_4 = \Delta x\big(f(0.5) + f(1.0) + f(1.5) + f(2.0)\big) = 0.5\,(0.25 + 1 + 2.25 + 4) = 0.5\,(7.5) = 3.75 .
$$

The exact value is $\int_0^2 x^2\,dx = \tfrac{x^3}{3}\big|_0^2 = \tfrac{8}{3} \approx 2.667$. The coarse sum overestimates; refining to $n = 100$ gives $\approx 2.69$, and $n \to \infty$ yields exactly $\tfrac83$.

```python
import sympy as sp

x = sp.symbols("x")
f = x**2
left_end, right_end = 0, 2

def right_riemann_sum(subinterval_count):
    width = sp.Rational(right_end - left_end, subinterval_count)
    sample_points = [left_end + width * index for index in range(1, subinterval_count + 1)]
    return width * sum(f.subs(x, point) for point in sample_points)

coarse_sum = right_riemann_sum(4)
fine_sum = right_riemann_sum(100)
exact_integral = sp.integrate(f, (x, left_end, right_end))

print("R_4   =", coarse_sum, "=", float(coarse_sum))
print("R_100 =", float(fine_sum))
print("exact =", exact_integral, "=", float(exact_integral))
```

**Equivalent CasADi implementation:**

The same right-endpoint sum, built as a CasADi expression and evaluated numerically, approaches the exact integral as the partition refines.

```python
import casadi as ca

left_end, right_end = 0.0, 2.0

def casadi_right_sum(subinterval_count):
    width = (right_end - left_end) / subinterval_count
    sample_points = [left_end + width * index for index in range(1, subinterval_count + 1)]
    return width * sum(point**2 for point in sample_points)

print("R_4   =", float(ca.DM(casadi_right_sum(4))))
print("R_100 =", float(ca.DM(casadi_right_sum(100))))
```
