### 1.1.8.3. Passive Transformation


$$
\begin{aligned}
\begin{pmatrix}
\text{Components of} \\
\text{same vector} \\
\text{in new system}
\end{pmatrix}
&=
\begin{pmatrix}
\text{Inverse} \\
\text{transformation} \\
\text{matrix}
\end{pmatrix}
\begin{pmatrix}
\text{Components of} \\
\text{vector in} \\
\text{original system}
\end{pmatrix},
\end{aligned}
$$

$$
\begin{aligned}
R_{\mathrm{passive}}(\theta)
&=
\begin{bmatrix}
\cos \theta & \sin \theta \\
-\sin \theta & \cos \theta
\end{bmatrix}
=
R_{\mathrm{direct}}(\theta)^{-1}
\end{aligned}
$$

**Explanation:**

A passive transformation changes the coordinate system while the geometric vector remains fixed. The basis vectors move by the direct transformation matrix, so the component column must move by the inverse matrix to preserve the same vector. For orthogonal rotations, the inverse is the transpose, which makes the passive and direct matrices easy to compare.

**Example:**

For $\theta=150^\circ$,

$$
P=\begin{bmatrix}\cos\theta&\sin\theta\\-\sin\theta&\cos\theta\end{bmatrix},
\qquad
D=\begin{bmatrix}\cos\theta&-\sin\theta\\\sin\theta&\cos\theta\end{bmatrix}.
$$

Multiplying them gives

$$
PD=
\begin{bmatrix}
\cos^2\theta+\sin^2\theta&-\cos\theta\sin\theta+\sin\theta\cos\theta\\
-\sin\theta\cos\theta+\cos\theta\sin\theta&\sin^2\theta+\cos^2\theta
\end{bmatrix}
=
\begin{bmatrix}1&0\\0&1\end{bmatrix}.
$$

So the passive and direct matrices undo each other.


```python
import sympy as sp

rotation_angle = sp.rad(150)
passive_transformation_matrix = sp.Matrix([
    [sp.cos(rotation_angle), sp.sin(rotation_angle)],
    [-sp.sin(rotation_angle), sp.cos(rotation_angle)],
])
direct_transformation_matrix = sp.Matrix([
    [sp.cos(rotation_angle), -sp.sin(rotation_angle)],
    [sp.sin(rotation_angle), sp.cos(rotation_angle)],
])
original_vector_components = sp.Matrix([5, 3])
transformed_vector_components = passive_transformation_matrix * original_vector_components
matrix_product = sp.simplify(passive_transformation_matrix * direct_transformation_matrix)

print("transformed_vector_components =")
sp.pprint(transformed_vector_components)
print("passive_transformation_matrix @ direct_transformation_matrix =")
sp.pprint(matrix_product)
```
