### A1.2.3. Observer Pattern

> *Define a one-to-many dependency between objects so that when one object changes state, all its dependents are notified and updated automatically.*
>
> â€” Gang of Four

**Explanation:**

The **Observer** pattern establishes a subscription mechanism where a *subject* (publisher) maintains a list of *observers* (subscribers) and notifies them whenever its state changes. Observers register and unregister themselves dynamically.

The subject knows nothing about the concrete observer classes â€” it interacts only through the observer interface. This decouples the source of events from the consumers of events.

**Properties:**

- Observers can be added or removed at runtime.
- The subject broadcasts to all registered observers without knowing their types.
- Each observer decides independently how to react to the notification.

**Example:**

A weather station (subject) measures temperature. Multiple display units (observers) â€” a current-conditions display, a statistics display â€” subscribe to the station. When the temperature changes, all displays update automatically.

````python
from abc import ABC, abstractmethod


class Observer(ABC):
    @abstractmethod
    def update(self, temperature):
        pass


class WeatherStation:
    def __init__(self):
        self._observers = []
        self._temperature = 0.0

    def register(self, observer):
        self._observers.append(observer)

    def unregister(self, observer):
        self._observers.remove(observer)

    def set_temperature(self, temperature):
        self._temperature = temperature
        self._notify()

    def _notify(self):
        for observer in self._observers:
            observer.update(self._temperature)


class CurrentConditionsDisplay(Observer):
    def update(self, temperature):
        print(f"Current conditions: {temperature}Â°C")


class StatisticsDisplay(Observer):
    def __init__(self):
        self._readings = []

    def update(self, temperature):
        self._readings.append(temperature)
        average = sum(self._readings) / len(self._readings)
        print(f"Statistics: avg={average:.1f}Â°C, readings={len(self._readings)}")


station = WeatherStation()
current_display = CurrentConditionsDisplay()
stats_display = StatisticsDisplay()

station.register(current_display)
station.register(stats_display)

station.set_temperature(22.5)
station.set_temperature(25.0)
station.set_temperature(19.3)
````

**Python Library â€” `blinker`:**

The [`blinker`](https://pypi.org/project/blinker/) library provides a fast, simple signal/event system used internally by Flask and Django. A `Signal` object acts as the subject; callables connect to it as observers.

```python
from blinker import signal

temperature_changed = signal("temperature-changed")

def log_temperature(sender, temperature):
    print(f"Logged: {temperature}Â°C from {sender}")

temperature_changed.connect(log_temperature)
temperature_changed.send("station-1", temperature=22.5)
```

`blinker` handles subscriber lifecycle, weak references, and named signals â€” replacing the manual observer list management shown above.
