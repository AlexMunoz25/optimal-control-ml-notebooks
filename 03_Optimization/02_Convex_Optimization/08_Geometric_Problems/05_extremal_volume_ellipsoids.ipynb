### 3.2.8.5. Extremal Volume Ellipsoids


$$\mathcal E=\{x_c+Au\mid\|u\|_2\leq1\}$$


**Explanation:**

Extremal ellipsoid problems approximate or cover [convex sets](../02_Convex_Sets/01_affine_sets_and_convex_sets.ipynb) with ellipsoids. The semiaxis lengths are determined by the shape matrix, and volume depends on their product. Feasibility is containment or inclusion, while optimality minimizes or maximizes volume.

The vector $x_c$ is the ellipsoid center, $A$ shapes the ellipsoid, and $u$ ranges over the unit Euclidean [norm](../../../01_Foundations/01_Linear_Algebra/02_Vector/06_vector_norm.ipynb) ball. The set is the image of that unit ball under an [affine map](../../../01_Foundations/01_Linear_Algebra/06_Linear_Transformations/08_affine_transformation.ipynb).

**Properties:**
- [Eigenvalues](../../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/08_eigenvalue.ipynb) of the shape matrix determine squared semiaxis lengths.
- Volume scales with the [determinant](../../../01_Foundations/01_Linear_Algebra/03_Matrix/10_matrix_determinant.ipynb).
- Ellipsoid problems become convex under standard parametrizations.

**Intuition:**

The example reads axis lengths from a diagonal ellipsoid shape matrix.


**Numerical Example:**

Let
$$
A=\begin{bmatrix}2&0\\0&1\end{bmatrix}.
$$

The ellipsoid $\{Au\mid\|u\|_2\leq1\}$ has shape matrix
$$
AA^T=\begin{bmatrix}4&0\\0&1\end{bmatrix}.
$$

The eigenvalues are $4$ and $1$, so the semiaxis lengths are
$$
\sqrt4=2,\qquad \sqrt1=1.
$$


```python
import sympy as sp

shape_matrix = sp.diag(4, 1)
axis_lengths = [sp.sqrt(eigenvalue) for eigenvalue in shape_matrix.eigenvals()]

shape_matrix, axis_lengths
```
