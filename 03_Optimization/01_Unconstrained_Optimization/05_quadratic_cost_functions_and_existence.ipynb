### 3.1.5. Quadratic Cost Functions and Existence


$$
\begin{aligned}
f(x) &= \frac{1}{2}x^{\mathsf T}Qx - b^{\mathsf T}x + c, \\
\nabla f(x) &= Qx-b, \\
Q \succ 0 &\Longrightarrow x^* = Q^{-1}b\ \text{is the unique global minimum}.
\end{aligned}
$$


**Explanation:**

Quadratic cost functions are the simplest nonlinear models that still contain the main geometry of unconstrained optimization. Their gradients are affine, their Hessians are constant, and the definiteness of the matrix $Q$ largely determines the behavior of the problem. When $Q$ is [positive definite](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/17_positive_definite_matrix.ipynb), the stationary equation $Qx=b$ has the unique solution $x^*=Q^{-1}b$, using the [matrix inverse](../../01_Foundations/01_Linear_Algebra/03_Matrix/11_matrix_inverse.ipynb), and this point is the unique global minimum. When $Q$ is singular or indefinite, existence and uniqueness require more care, and the [eigenvalues](../../01_Foundations/01_Linear_Algebra/07_Theoretical_Linear_Algebra/08_eigenvalue.ipynb) of $Q$ reveal what can go wrong.

The display writes a quadratic objective with [matrix](../../01_Foundations/01_Linear_Algebra/03_Matrix/01_matrix_definition.ipynb) $Q$, vector $b$, scalar $c$, and decision vector $x$. The term $x^{\mathsf T}Qx$ is a quadratic form; when $Q$ is positive definite, the linear equation $Qx=b$ has the unique minimizer $x^*=Q^{-1}b$, using a [matrix inverse](../../01_Foundations/01_Linear_Algebra/03_Matrix/11_matrix_inverse.ipynb).

Quadratics deserve special attention because later convergence-rate analysis repeatedly reduces smooth problems to local quadratic approximations. In that sense, a quadratic is not merely an example; it is the model problem behind gradient, Newton, conjugate-gradient, and quasi-Newton behavior.

The Weber-frame figure shows that optimality and existence also arise from physical equilibrium interpretations, not only from algebraic formulas. **Numerical Example:**

The positive definite case in the quadratic-cost figure has one elliptical bowl. Let
$$
Q=\begin{bmatrix}4&1\\1&2\end{bmatrix},\qquad b=\begin{bmatrix}1\\3\end{bmatrix},\qquad f(x)=\frac{1}{2}x^{\mathsf T}Qx-b^{\mathsf T}x.
$$

The stationary equation $Qx=b$ is
$$
\begin{aligned}
4x_1+x_2&=1,\\
x_1+2x_2&=3.
\end{aligned}
$$

From the first equation, $x_2=1-4x_1$. Substitute into the second equation:
$$
x_1+2(1-4x_1)=3
\quad\Longrightarrow\quad
-7x_1=1
\quad\Longrightarrow\quad
x_1=-\frac{1}{7},\qquad x_2=\frac{11}{7}.
$$

At the minimizer, $Qx^*=b$, so
$$
f(x^*)=\frac{1}{2}x^{*\mathsf T}b-b^{\mathsf T}x^*
=-\frac{1}{2}b^{\mathsf T}x^*
=-\frac{1}{2}\left(-\frac{1}{7}+3\cdot\frac{11}{7}\right)
=-\frac{16}{7}.
$$


```python
import numpy as np

Q = np.array([[4.0, 1.0], [1.0, 2.0]])
linear_term = np.array([1.0, 3.0])
minimizer = np.linalg.solve(Q, linear_term)
minimum_value = 0.5 * minimizer @ Q @ minimizer - linear_term @ minimizer

minimizer, minimum_value
```
