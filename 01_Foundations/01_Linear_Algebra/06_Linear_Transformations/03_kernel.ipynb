### 1.1.6.3. Kernel


$$
\operatorname{Ker}(T) =
\left\{
\mathbf{v} \in V \mid T(\mathbf{v}) = \mathbf{0}
\right\}
$$


**Explanation:**

The kernel is the set of inputs that a linear transformation sends to the zero vector. It is the transformation analogue of the [null space](../05_Geometrical_Aspects_of_Linear_Algebra/16_matrix_fundamental_spaces.ipynb), so it identifies directions that are lost under the map. Its role in reversibility is developed in [invertible linear transformations](./07_invertible_linear_transformation.ipynb).

**Example:**

For

$$
T(x, y, z) =
\begin{bmatrix}
x \\
z
\end{bmatrix},
$$

the kernel is

$$
\operatorname{Ker}(T) =
\operatorname{span}
\left\{
\begin{bmatrix}
0 \\
1 \\
0
\end{bmatrix}
\right\}.
$$


```python
import sympy as sp

transformation_matrix = sp.Matrix([[1, 0, 0], [0, 0, 1]])
kernel_basis = transformation_matrix.nullspace()

print("kernel_basis =", kernel_basis)
```
