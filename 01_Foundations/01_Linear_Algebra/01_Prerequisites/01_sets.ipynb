**Explanation:**

Mathematicians classify the different kinds of number-like objects into categories called **sets**:

- **The natural numbers:** $\mathbb{N} = \{0, 1, 2, 3, 4, 5, 6, 7, \ldots\}$
- **The integers:** $\mathbb{Z} = \{\ldots, -3, -2, -1, 0, 1, 2, 3, \ldots\}$
- **The rational numbers:** $\mathbb{Q} = \left\{\frac{5}{3}, \frac{22}{7}, 1.5, 0.125, -7, \ldots\right\}$
- **The real numbers:** $\mathbb{R} = \{-1, 0, 1, \sqrt{2}, e, \pi, 4.94\ldots, \ldots\}$
- **The complex numbers:** $\mathbb{C} = \{-1, 0, 1, i, 1+i, 2+3i, \ldots\}$

**Definitions:**

- *Set*: a collection of mathematical objects.
- $s \in S$: "$s$ is an element of $S$."
- $\varnothing$: the *empty set* contains no elements.

**Set operations:**

- $S \cup T$: the *union* — elements in either $S$ or $T$
- $S \cap T$: the *intersection* — elements in both $S$ and $T$
- $S \setminus T$: the *set difference* — elements of $S$ not in $T$

**Set relations:**

- $S \subset T$: *strict subset*
- $S \subseteq T$: *subset or equal*

The coordinate values in [The Cartesian Plane](./02_cartesian_plane.ipynb) are chosen from these sets. Later objects such as [Vectors](../02_Vector/01_vector_definition.ipynb) and [Matrices](../03_Matrix/01_matrix_definition.ipynb) use entries drawn from the same number sets.

**Example:**

Let

$$
S = \{1, 2, 3\}, \quad T = \{3, 4\}.
$$

Then

$$
S \cup T = \{1, 2, 3, 4\}, \quad S \cap T = \{3\}, \quad S \setminus T = \{1, 2\}.
$$


```python
set_s = {1, 2, 3}
set_t = {3, 4}

print("S =", set_s)
print("T =", set_t)
print("S ∪ T =", set_s | set_t)
print("S ∩ T =", set_s & set_t)
print("S \\ T =", set_s - set_t)
print("T \\ S =", set_t - set_s)
print("S △ T =", set_s ^ set_t)
print("S ⊂ T =", set_s < set_t)
print("S ⊆ T =", set_s <= set_t)
print("S ⊃ T =", set_s > set_t)
print("S ⊇ T =", set_s >= set_t)
print("S ∩ T = ∅:", set_s.isdisjoint(set_t))
```
