### 1.3.3.5. Integration by Parts

$$
\int u\,dv = uv - \int v\,du,
\qquad\text{equivalently}\qquad
\int f(x)\,g'(x)\,dx = f(x)g(x) - \int f'(x)\,g(x)\,dx .
$$

**Explanation:**

Integration by parts is the [product rule](../02_Single_Variable_Differentiation/02_differentiation_rules.ipynb) read backwards: integrating $(fg)' = f'g + fg'$ rearranges to $\int fg' = fg - \int f'g$. It trades one integral for another, and is useful when the new integrand $f'g$ is simpler than the original — for instance when $f$ is a polynomial that differentiates toward a constant. The same "move the derivative onto the other factor" manipulation reappears as the [adjoint / integration-by-parts step](../06_Multivariable_Integration_and_Vector_Calculus/07_divergence_theorem_and_integration_by_parts.ipynb) that derives optimality conditions in variational and optimal-control problems.

**Properties:**
- Choose $u$ to be the factor whose derivative simplifies and $dv$ the factor that is easy to integrate.
- Definite form: $\displaystyle\int_{a}^{b} f g'\,dx = \big[fg\big]_{a}^{b} - \int_{a}^{b} f' g\,dx$.

**Numerical Example:**

Evaluate $\int x\,e^{x}\,dx$. Choose $u = x$ (so $du = dx$) and $dv = e^{x}\,dx$ (so $v = e^{x}$):

$$
\int x\,e^{x}\,dx = uv - \int v\,du = x e^{x} - \int e^{x}\,dx = x e^{x} - e^{x} + C = (x - 1)e^{x} + C .
$$

Check: $\dfrac{d}{dx}\big[(x-1)e^x\big] = e^x + (x-1)e^x = x e^x$. The polynomial factor $x$ differentiated to $1$, simplifying the remaining integral.

```python
import sympy as sp

x = sp.symbols("x")
f = x
dv_antiderivative = sp.exp(x)

by_parts = f * dv_antiderivative - sp.integrate(sp.diff(f, x) * dv_antiderivative, x)
direct_result = sp.integrate(x * sp.exp(x), x)
check = sp.simplify(sp.diff(by_parts, x))

print("by parts: uv - ∫v du =", sp.simplify(by_parts), "+ C")
print("direct sympy        :", direct_result, "+ C")
print("d/dx of result       :", check)
```
